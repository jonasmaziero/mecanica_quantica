# Relações de incerteza em Mecânica Quântica
Vimos que a existência de observáveis incompatíveis, que não comutam, possibilita a preparação de estados de superposição que tem a eles associada uma "incerteza quântica" com relação a medidas futuras. Aqui usaremos o formalismo cinemático quântico para obter relações matemáticas gerais que fornecem limitações para o produto ou soma das "incertezas" (desvios padrões) sobre dois observáveis. Esse texto é baseado em [arXiv:1705.09139].

## Relação de incerteza de Heisenberg-Robertson-Schrödinger (RIHRS)
Aqui usaremos como uma das principais ferramentas a desigualdade de Cauchy-Scharz (DCS):
\begin{equation}
\langle\psi|\psi\rangle\langle\phi|\phi\rangle\ge|\langle\psi|\phi\rangle|^{2},
\end{equation}
como igualdade obtida se e somente se $|\psi\rangle$ e $|\phi\rangle$ são colineares.<br>
__Exercício:__ Prove a DCS.

Considere $A$ e $B$ dois observáveis de um sistema físico preparado no estado $|\xi\rangle$. Usaremos $\langle X\rangle=\langle\xi|X|\xi\rangle$ para denotar o valor médio de um operador qualquer $X$ e $\mathbb{I}$ para o operador identidade $\mathcal{H}$. Vamos definir ainda os seguintes vetores:
\begin{equation}
|\psi\rangle=(A-\langle A\rangle\mathbb{I})|\xi\rangle\mbox{ e }|\phi\rangle=(B-\langle B\rangle\mathbb{I})|\xi\rangle, \label{eq:states}
\end{equation}
que substituiremos na DCS. Notemos que 
\begin{align}
\langle\psi|\psi\rangle & = \langle\xi|(A-\langle A\rangle\mathbb{I})(A-\langle A\rangle\mathbb{I})|\xi\rangle \\
                        & = \langle\xi|(A-\langle A\rangle\mathbb{I})^{2}|\xi\rangle \\
                        & = \mathrm{Var}(A) \\ 
\langle\phi|\phi\rangle & = \mathrm{Var}(B),
\end{align}
com $\mathrm{Var}(X)=\langle(X-\langle X\rangle\mathbb{I})^{2}\rangle$
sendo a variância de $X$. Verifica-se que
\begin{eqnarray}
\langle\psi|\phi\rangle & = & \langle AB-\langle A\rangle\langle B\rangle\mathbb{I}\rangle\\
 & = & 2^{-1}\langle\{A,B\}-2\langle A\rangle\langle B\rangle\mathbb{I}\rangle+2^{-1}\langle[A,B]\rangle \\
 & = & \mathrm{Re} \langle\psi|\phi\rangle + i \mathrm{Im} \langle\psi|\phi\rangle \\
 & = & 2^{-1}(\langle AB\rangle -\langle A\rangle\langle B\rangle + \langle BA\rangle - \langle A\rangle\langle B\rangle) + 2^{-1}\langle[A,B]\rangle
\end{eqnarray}
onde
\begin{equation}
[A,B]=AB-BA\mbox{ e }\{A,B\}=AB+BA
\end{equation}
são, respectivamente, o comutador e anti-comutador de $A$ e $B$. Note que $AB = 2^{-1}(\{A,B\}+[A,B])$. Como $\{A,B\}$ e $[A,B]$ são, respectivamente, operadores Hermitianos e anti-Hermitianos, seus valores médios são, respectivamente, números reais e imaginários puros. Assim, considerando que 
\begin{equation}
|\langle\psi|\phi\rangle|^{2}=(\mathrm{Re}\langle\psi|\phi\rangle)^{2}+(\mathrm{Im}\langle\psi|\phi\rangle)^{2},
\end{equation}
depois de algumas manipulações algébricas obteremos a RIHRS:
\begin{equation}
\mathrm{Var}(A)\mathrm{Var}(B)\ge (\mathrm{CovQ}(A,B))^{2}+2^{-2}|\langle[A,B]\rangle|^{2},\label{eq:RIHRS}
\end{equation}
onde
\begin{equation}
\mathrm{CovQ}(A,B)=2^{-1}(\mathrm{Cov}(A,B)+\mathrm{Cov}(B,A))
\end{equation}
é a covariância quântica, com $\mathrm{Cov}(X,Y)=\langle XY\rangle-\langle X\rangle\langle Y\rangle$
sendo a covariância entre os observáveis $X$ e $Y$.
Se $A$ e $B$ são observáveis compatíveis, i.e., se
$[A,B]=\mathbb{0}$, teremos que $\mathrm{CovQ}(A,B)=\mathrm{Cov}(A,B)$.

Vamos analizar agora o _problema da possível trivialidade_ da RIHRS. Vamos assumir que o sistema está preparado em um estado que é autovetor do observável $A$, i.e., $|\xi\rangle=|a\rangle$ com $A|a\rangle=a|a\rangle$ e $a\in\mathbb{R}$. Neste caso pode-se verificar que (__exercício__): $\mathrm{Var}(A)=\mathrm{CovQ}(A,B)=\langle[A,B]\rangle=0$. E com isso a RIHRS nos retorna 
\begin{equation}
0\mathrm{Var}(\hat{B})\ge0.
\end{equation}
Por conseguinte, neste caso, a RIHRS não fornece informação alguma sobre a possível incompatibilidade de $A$ e $B$.

### Relação de incerteza de Heisenberg (RIH)
Ignorando-se o termo relacionado à covariância quântica na RIHRS e tomando a raiz quadrada da expressão que resta, obtemos a _relação de incerteza de Robertson_:
\begin{equation}
\sigma_{A}\sigma_{B} \ge 2^{-1}|\langle[A,B]\rangle|,
\end{equation}
que quando aplicada a posição e momento linear é uma das _relações de incerteza de Heisenberg_:
\begin{equation}
\sigma_{X}\sigma_{P} \ge 2^{-1}|\langle[X,P]\rangle|.
\end{equation}

## Relação de incerteza de Maccone-Pati (RIMP)
Essa relação de incerteza lê-se:
\begin{equation}
\mathrm{Var}(A)+\mathrm{Var}(B) \ge \frac{1}{2}|\langle\xi|(A\pm B)|\xi_{\perp}\rangle|^{2},
\end{equation}
com $|\xi_{\perp}\rangle$ sendo qualquer vetor normalizado e ortogonal a $|\xi\rangle$ e o sinal é escolhido de forma a maximizar o limite inferior.

Para provar a RIMP, começamos com a regra dos paralelogramos:
\begin{equation}
2(|||\psi\rangle||^{2}+|||\phi\rangle||^{2}) = |||\psi\rangle|+|\phi\rangle||^{2} + |||\psi\rangle-|\phi\rangle||^{2}.
\end{equation}

__Exercício:__ Verifique a validade geral dessa regra.

Como fizemos no caso da RIHRS, começamos definindo $|\psi\rangle=(A-\langle A\rangle\mathbb{I})|\xi\rangle \Rightarrow |||\psi\rangle||^{2}=\mathrm{Var}(A)$ e $|\phi\rangle=(B-\langle B\rangle\mathbb{I})|\xi\rangle \Rightarrow |||\phi\rangle||^{2}=\mathrm{Var}(B)$ para escrever
\begin{align}
\mathrm{Var}(A)+\mathrm{Var}(B) & =  \frac{1}{2}\left(|||\psi\rangle|+|\phi\rangle||^{2} + |||\psi\rangle-|\phi\rangle||^{2}\right) \\
                                & \ge  \frac{1}{2}|||\psi\rangle|\pm|\phi\rangle||^{2} \\
                                & =  \frac{1}{2}(\langle\psi|\pm\langle\phi|)(|\psi\rangle\pm|\phi\rangle)\langle\xi_{\perp}|\xi_{\perp}\rangle \\
                                & \ge  \frac{1}{2}|(\langle\psi|\pm\langle\phi|)|\xi_{\perp}\rangle|^{2} \\
                                & =  \frac{1}{2}|(\langle\xi|(A-\langle A\rangle\mathbb{I})\pm\langle\xi|(B-\langle B\rangle\mathbb{I}))|\xi_{\perp}\rangle|^{2} \\
                                & =  \frac{1}{2}|\langle\xi|(A\pm B)|\xi_{\perp}\rangle - (\langle A\rangle\pm \langle B\rangle)\langle\xi|\xi_{\perp}\rangle|^{2} \\
                                & =  \frac{1}{2}|\langle\xi|(A\pm B)|\xi_{\perp}\rangle|^{2}.
\end{align}

### Exemplo: 
#### Relação de incerteza entre campo elétrico (fase) e o número de fótons (intensidade) do campo eletromagnético quantizado
Um modo de oscilação do campo eletromagnético dentro de uma cavidade, quando quantizado, é descrito pelo Hamiltoniano do oscilador harmônico quântico unidimensional:
\begin{equation}
H = \hbar\omega(N+\mathbb{I}/2),
\end{equation}
onde $N=a^{\dagger}a$ é o operador número e $a^{\dagger}$ e $a$ são os operadores de criação e aniquilação, respectivamente.

Da solução analítica para os autovalores de $H$, sabe-se que
\begin{align}
& N|n\rangle = n|n\rangle, \\
& a|n\rangle = \sqrt{n}|n-1\rangle, \\
& a^{\dagger}|n\rangle =  \sqrt{n+1}|n+1\rangle,
\end{align}
onde $n = 0, 1, 2, \cdots, \infty$. 

O operador Hermitiano que descreve o campo elétrico é
\begin{equation}
E_{x} = E_{0x}\sin kz (a+a^{\dagger}) =: E(a+a^{\dagger})\propto \text{ quadratura } X,
\end{equation}
com $E_{0x}$ e $k$ escalares reais. 

Vamos considerar o campo preparado no seguinte estado de superposição:
\begin{equation}
|\xi\rangle = \cos\theta|0\rangle+\sin\theta|1\rangle,
\end{equation}
e vamos calcular as variâncias de $N$ e de $E_{x}$ e os limites inferiores dados pelas RIHRS e RIMP para essas quantidades.

#### Variâncias
Para o operador número:
\begin{align}
\langle N\rangle_{\xi} & = \langle\xi|N(\cos\theta|0\rangle+\sin\theta|1\rangle) 
= \langle\xi|(\cos\theta N|0\rangle+\sin\theta N|1\rangle) 
= (\cos\theta\langle 0|+\sin\theta\langle 1|)(\cos\theta (0)|0\rangle+\sin\theta (1)|1\rangle) \\ 
                       & = \sin^{2}\theta = \langle N^{2}\rangle_{\xi}, \\
\mathrm{Var}(N) & = \langle N^{2}\rangle_{\xi} - \langle N\rangle_{\xi}^{2} = \sin^{2}\theta\cos^{2}\theta.
\end{align}

Note que $|\xi\rangle$ não é autovetor de $E_{x}$ para nenhum valor de $\theta$ pois
\begin{equation}
(a+a^{\dagger})|\xi\rangle = \sin\theta|0\rangle + \cos\theta|1\rangle + \sqrt{2}\sin\theta|2\rangle,
\end{equation}
ou seja, $\sigma_{E_{x}}>0$. Aqui
\begin{align}
\langle E_{x}\rangle_{\xi} & = E\langle\xi|(a+a^{\dagger})|\xi\rangle \\
                           & = E(\cos\theta\langle 0|+\sin\theta\langle 1|)(\sin\theta|0\rangle + \cos\theta|1\rangle + \sqrt{2}\sin\theta|2\rangle) \\
                           & = E2\sin\theta\cos\theta \\ 
                           & = E\sin 2\theta, \\
\langle E_{x}^{2}\rangle_{\xi} & = E^{2} \langle\xi|(a+a^{\dagger})(a+a^{\dagger})|\xi\rangle \\
                               & = E^{2} \langle\xi|(a+a^{\dagger})(\sin\theta|0\rangle + \cos\theta|1\rangle + \sqrt{2}\sin\theta|2\rangle) \\
                               & = E^{2} (\cos\theta\langle 0|+\sin\theta\langle 1|)(0 + \cos\theta|0\rangle + 2\sin\theta|1\rangle + \sin\theta|1\rangle + \sqrt{2}\cos\theta|2\rangle + \sqrt{6}\sin\theta|3\rangle) \\
                               & = E^{2} (\cos^{2}\theta + 3\sin^{2}\theta) = E^{2} (1 + 2\sin^{2}\theta) = E^{2} (1 + 2\frac{1-\cos 2\theta}{2}) \\
                               & = E^{2} (2-\cos 2\theta), \\
\mathrm{Var}(E_{x}) & = \langle E_{x}^{2}\rangle_{\xi} - \langle E_{x}\rangle_{\xi}^{2} \\
                    & = E^{2}(2 - \cos2\theta - \sin^{2}2\theta).
\end{align}

#### Limites inferiores
Usaremos
\begin{equation}
[a,a^{\dagger}]=\mathbb{I}
\end{equation}
para obter o comutador de $N$ com $E_{x}$:
\begin{align}
[N,E_{x}] & = E([a^{\dagger}a,a] + [a^{\dagger}a,a^{\dagger}]) \\
          & = E(a^{\dagger}[a,a] + [a^{\dagger},a]a + a^{\dagger}[a,a^{\dagger}] + [a,a^{\dagger}]a^{\dagger}) \\
          & = E(-a + a^{\dagger}) \\
          & \propto \text{ quadratura } P.
\end{align}

Valor médio do comutador:
\begin{align}
\langle[N,E_{x}]\rangle_{\xi} & = E\langle\xi|(-a + a^{\dagger})(\cos\theta|0\rangle+\sin\theta|1\rangle) \\
                              & = E(\cos\theta\langle0|+\sin\theta\langle1A)(0-\sin\theta|0\rangle+\cos\theta|1\rangle+\sqrt{2}\sin\theta|2\rangle) \\
                              & = E(-\sin\theta\cos\theta + \sin\theta\cos\theta) \\
                              & = 0.
\end{align}

Para o anticomutador teremos
\begin{align}
\{N,E_{x}\} & = E(\{a^{\dagger}a,a\} + \{a^{\dagger}a,a^{\dagger}\}) \\
            & = E(a^{\dagger}aa + aa^{\dagger}a + a^{\dagger}aa^{\dagger} + a^{\dagger}a^{\dagger}a).
\end{align}
Notemos que
\begin{align}
a^{\dagger}aa|\xi\rangle & = 0, \\
aa^{\dagger}a|\xi\rangle & = \sin\theta|0\rangle, \\
a^{\dagger}aa^{\dagger}|\xi\rangle & = \cos\theta|1\rangle + 2\sqrt{2}\sin\theta|2\rangle, \\
a^{\dagger}a^{\dagger}a & = \sqrt{2}\sin\theta|2\rangle.
\end{align}
Assim
\begin{align}
\langle\{N,E_{x}\}\rangle_{\xi} & = E\langle\xi(a^{\dagger}aa + aa^{\dagger}a + a^{\dagger}aa^{\dagger} + a^{\dagger}a^{\dagger}a)|\xi\rangle \\
                                & = E(\cos\theta\langle\xi|+\sin\theta\langle1|)i(0 + \sin\theta|0\rangle + \cos\theta|1\rangle + 2\sqrt{2}\sin\theta|2\rangle + \sqrt{2}\sin\theta|2\rangle)|\xi\rangle \\
                                & = E(2\sin\theta\cos\theta) = E\sin2\theta.
\end{align}

Covariância quântica
\begin{align}
\mathrm{CovQ}(N,E_{x})_{\xi} & = (\mathrm{Cov}(N,E_{x})_{\xi} + \mathrm{Cov}(E_{x},N)_{\xi})/2 \\
                       & = \left(\langle NE_{x}\rangle_{\xi} - \langle N\rangle_{\xi}\langle E_{x}\rangle_{\xi} + \langle E_{x}N\rangle_{\xi} - \langle N\rangle_{\xi}\langle E_{x}\rangle_{\xi} \right)/2 \\
                       & = 2^{-1}\langle\{N,E_{x}\}\rangle_{\xi} - \langle N\rangle_{\xi}\langle E_{x}\rangle_{\xi} \\
                       & = 2^{-1}E\sin2\theta - \sin^{2}\theta E\sin2\theta \\
                       & = E\sin2\theta(2^{-1} - \sin^{2}\theta).
\end{align}

Limite inferior para a RIHRS:
\begin{align}
L_{hrs} & = (\mathrm{CovQ}(N,E_{x}))^{2} + 2^{-2}|\langle[N,E_{x}]\rangle|^{2} \\
        & = \left(E\sin2\theta(2^{-1} - \sin^{2}\theta)\right)^{2} + 0.
\end{align}

Limite inferior para a RIMP:
\begin{align}
L_{mp} & = \frac{1}{2}|\langle\xi_{\perp}|(N\pm E_{x})|\xi\rangle|^{2} \\
        & = \frac{1}{2}|\langle\xi_{\perp}|(a^{\dagger}a\pm E(a+a^{\dagger}))(\cos\theta|0\rangle+\sin\theta|1\rangle)|^{2} \\
        & = \frac{1}{2}|\langle\xi_{\perp}|(0+\sin\theta|1\rangle\pm E(0+\sin\theta|0\rangle +\cos\theta|1\rangle+\sqrt{2}\sin\theta|2\rangle))|^{2} \\
        & = \frac{1}{2}|\langle\xi_{\perp}|((\sin\theta \pm E \cos\theta)|1\rangle \pm E(\sin\theta|0\rangle +\sqrt{2}\sin\theta|2\rangle))|^{2} \\
\end{align}
A escolha mais simples para $|\xi_{\perp}\rangle$ seria $|2\rangle$. Mas nesse caso $L_{mp}=\sin^{2}\theta$ será trivial para $\theta=0,\pi$. Uma possibilidade mais geral seria
\begin{equation}
|\xi_{\perp}\rangle = \frac{\sin\theta}{\sqrt{2}}|0\rangle - \frac{\cos\theta}{\sqrt{2}}|1\rangle + \frac{1}{\sqrt{2}}|2\rangle.
\end{equation}
Nesse caso teremos
\begin{align}
L_{mp} & = \frac{1}{2}\left| \pm\frac{E\sin^{2}\theta}{\sqrt{2}} -\frac{\cos\theta}{\sqrt{2}}(\sin\theta\pm E\cos\theta) \pm\frac{E\sqrt{2}\sin\theta}{\sqrt{2}} \right|^{2} \\
       & = \vdots \\
       & = \left|\mp\frac{E\cos2\theta}{2} -\frac{\sin2\theta}{4} \pm \frac{E\sin\theta}{\sqrt{2}} \right|^{2}.
\end{align}
Ainda temos pontos de trivialidade. Outro $|\xi_{\perp}\rangle$? <br>
Por outro lado, podemos olhar para o outro limite inferior da RIMP (ver [arXiv:1705.09139]):
\begin{align}
\tilde{L}_{mp} & = \pm i\langle[N,E_{x}]\rangle + |\langle\xi_{\perp}|(N\pm iE_{x})|\xi\rangle|^{2} \\
               & = \vdots \\
               & = \left| -\frac{\sin2\theta}{4} \pm iE\left(-\frac{\cos2\theta}{2} + \frac{\sin\theta}{\sqrt{2}} \right) \right|^{2} \\
               & = \frac{\sin^{2}2\theta}{16} + E^{2}\left(\frac{\cos2\theta}{2}+\frac{\sin\theta}{\sqrt{2}}\right)^{2}.
\end{align}
Aqui usamos o mesmo $|\xi_{\perp}\rangle$ que acima. Nesse caso não há problema de trivialidade para nenhum valor de $\theta$.

```python
N, th, dth = symbols('N theta dth')
N = 150
dth = (np.pi-0)/N
Nm = np.zeros(N)
VarN = np.zeros(N)
Em = np.zeros(N)
VarE = np.zeros(N)
Lhrs = np.zeros(N)
Lmp = np.zeros(N)
Lmpt = np.zeros(N)
thv = np.zeros(N)
th = -dth
for j in range(0,N):
    th += dth
    thv[j] = th
    Nm[j] = (sin(th))**2
    VarN[j] = abs(sin(th)*cos(th))
    Em[j] = sin(2*th)
    VarE[j] = sqrt(2 - cos(2*th) - (sin(2*th))**2)
    Lhrs[j] = abs(sin(2*th)*(0.5-(sin(th))**2))
    Lmp[j] = abs(+cos(2*th)/2 - sin(2*th)/4 - sin(th)/sqrt(2))
    Lmpt[j] = sqrt(sin(2*th)**2/16 + (-cos(2*th)/2 + sin(th)/sqrt(2))**2)
plt.plot(thv, Nm, label=r'$\langle N\rangle$')
plt.plot(thv, VarN, label=r'$\sigma_{N}$')
plt.plot(thv, Em, label=r'$\langle E_{x}\rangle/E$')
plt.plot(thv, VarE, label=r'$\sigma_{E_{x}}/E$')
plt.plot(thv, Lhrs, label=r'$\sqrt{L_{hrs}}/E$')
plt.plot(thv, Lmp, label=r'$\sqrt{L_{mp}}$, $E=1$')
plt.plot(thv, Lmpt, label=r'$\sqrt{\tilde{L}_{mp}}$, $E=1$')
plt.xlabel(r'$\theta$')
axes = plt.gca()
axes.set_xlim([0,np.pi])
axes.set_ylim([-1.02,1.8])
plt.legend()
plt.show()
```

## Relação de incerteza energia-tempo (RIET)
O tempo é só um parâmetro em MQ, não existe um operador Hermitiano que o represente. A seguir obteremos o que se chama de RIET, clarificando a que tempo essas relações se referem. Consideremos a eq. de Heisenberg,
\begin{equation}
i\hbar\partial_{t}O_{t}^{H} = [O_{t}^{H},H_{t}^{H}],
\end{equation}
e a relação de incerteza de Robertson:
\begin{equation}
\sigma_{O_{t}}\sigma_{H_{t}} \ge 2^{-1}|\langle[O_{t}^{H},H_{t}^{H}]\rangle|.
\end{equation}
Teremos assim que
\begin{align}
|\langle i\hbar\partial_{t}O_{t}^{H}\rangle| = \hbar|\partial_{t}\langle O_{t}\rangle| = |\langle[O_{t}^{H},H_{t}^{H}]\rangle| \le 2\sigma_{O_{t}}\sigma_{H_{t}}.
\end{align}
Se $O$ não é uma constante de movimento, i.e., se $\partial_{t}\langle O_{t}\rangle \ne 0$, podemos escrever a seguinte desigualdade:
\begin{equation}
\sigma_{H_{t}}\frac{\sigma_{O_{t}}}{|\partial_{t}\langle O_{t}\rangle|} \ge \frac{\hbar}{2}.
\end{equation}
Definiremos agora a seguinte escala de tempo
\begin{equation}
\Delta t_{O} := \frac{\sigma_{O_{t}}}{|\partial_{t}\langle O_{t}\rangle|},
\end{equation}
o que nos leva à RIET:
\begin{equation}
\sigma_{H_{t}}\Delta t_{O} \ge \frac{\hbar}{2}.
\end{equation}
Note que o tempo que aparece nessa expressão não é em geral aquele medido com relógio, mas sim a razão entre a largura da distribuição de probabilidades de $O$ no tempo $t$ e a taxa de mudança do centro dessa distribuição nesse instante.

### Limite de rapidez para mudanças de estados quânticos
Consideremos o sistema inicialmente preparado no estado $|\xi\rangle$. Vamos escrever o estado evoluído como
\begin{equation}
|\xi_{t}\rangle = \cos\theta_{t}|\xi\rangle + \sin\theta_{t}|\xi_{\perp}\rangle,
\end{equation}
com $\langle\xi|\xi_{\perp}\rangle=0$. Na RIET usamos o operador Hermitiano $O = |\xi\rangle\langle\xi|$ e faremos as contas na representação de Schrödinger considerando um Hamiltoniano independente do tempo:
\begin{align}
& \langle O^{2}_{t}\rangle = \langle O_{t}\rangle = \langle\xi_{t}|\xi\rangle\langle\xi|\xi_{t}| = \cos^{2}\theta_{t}, \\
& \sigma_{O_{t}} = \sqrt{\langle O^{2}_{t}\rangle - \langle O_{t}\rangle^{2}} = \sqrt{\cos^{2}\theta_{t}(1-\cos^{2}\theta_{t})} = |\sin\theta_{t}\cos\theta_{t}|, \\
& \partial_{t}\langle O_{t}\rangle = -2\cos\theta_{t}\sin\theta_{t}\partial_{t}\theta_{t}, \\
& \Delta t_{O} = \frac{|\sin\theta_{t}\cos\theta_{t}|}{|-2\cos\theta_{t}\sin\theta_{t}\partial_{t}\theta_{t}|} = \frac{1}{2|\partial_{t}\theta_{t}|}, \\
& \Delta t_{O} \ge \frac{\hbar}{2\sigma_{H_{t}}} \Rightarrow \frac{1}{2|\partial_{t}\theta_{t}|} \ge \frac{\hbar}{2\sigma_{H_{t}}} \Rightarrow \frac{\sigma_{H}}{\hbar} \ge |\partial_{t}\theta_{t}| \Rightarrow \frac{\sigma_{H}t}{\hbar} \ge \theta_{t} \Rightarrow t \ge \frac{\hbar\theta_{t}}{\sigma_{H}}.
\end{align}
Então o tempo necessário para obtermos a transição de $|\xi\rangle$ para o estado ortogonal $|\xi_{\perp}\rangle$ é
\begin{equation}
t_{\perp} \ge \frac{\hbar\pi}{2\sigma_{H}},
\end{equation}
ou seja, quanto maior a largura da distribuição de probabilidades do Hamiltoniano em um estado, mas rápido este estado pode mudar com o tempo.
