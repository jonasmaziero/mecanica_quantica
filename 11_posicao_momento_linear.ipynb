# Mecânica Quântica na base de posição

## Operador posição e função de onda, na base de posição
As três coordenadas espaciais que especificam a posição de uma partícula em relação a algum referencial são mensuráveis e devem portanto ser descritas, dentro do formalismo quântico, por operadores Hermitianos $X_{j},$ para $j=1,2,3$, tais que 
\begin{equation}
X_{j}|x_{j}\rangle=x_{j}|x_{j}\rangle
\end{equation}
com $x_{j}\in\mathbb{R}$ e
\begin{equation}
\langle x_{j}|x_{j}'\rangle=\delta(x_{j}-x_{j}').
\end{equation}
Note que acima estamos assumindo implicitamente e teoricamante que dois estados de posição infinitesimalmente próximos são distinguíveis, o que não é factível experimentalmente. Continuando com a teoria, como $X_{j}$ possui espectro contínuo, sua decomposição espectral será
\begin{equation}
X_{j} = \int dx_{j}x_{j}|x_{j}\rangle\langle x_{j}|.
\end{equation}

Como fizemos para outros observáveis, se fazemos medidas de posição para um sistema quântico preparado em um estado genérico $|\psi\rangle$, é conveniente escrevermos
\begin{equation}
|\psi\rangle = \int dx_{j}c_{x_{j}}|x_{j}\rangle.
\end{equation}
Os coeficientes nessa superposição são conhecidos como __função de onda__ na base de posição:
\begin{equation}
c_{x_{j}} = \langle x_{j}|\psi\rangle =: \psi(x_{j}).
\end{equation}

Para observáveis com espectro contínuo, existe uma complicação na interpretação da ortogonalidade entre autovetores correspondentes a autovalores diferentes. Isso se deve ao fato de que na prática só conseguimos distinguir experimentalmente entre valores cuja diferença é finita, e não infinitesimal. Tirando esse detalhe, a _interpretação estatística_ é a usual. A regra de Born nos diz que
\begin{equation}
|\psi(x_{j})|^{2}dx_{j}
\end{equation}
é a probabilidade de ao medirmos $X_{j}$ obtermos seu valor entre $x_{j}$ e $x_{j}+dx_{j}$, com $|\psi(x_{j})|^{2}$ sendo a densidade de probabilidade. Claro, como a partícula deve estar em algum lugar, devemos ter a _condição de normalização_:
\begin{equation}
\int_{-\infty}^{+\infty}|\psi(x_{j})|^{2}dx_{j} = 1.
\end{equation}

OBS. Aqui consideramos que não há emaranhamento entre as coordenadas do quanton. No entanto, definições análogas seguem para o caso geral, usando o vetor posição com $|\psi\rangle = \iiint dV c_{\vec{r}}|\vec{r}\rangle$, com $dV$ sendo um elemento de volume.

## Operador de translação espacial
Esse operador descreve, como o nome diz, mudanças do estado de posição. Para mudanças infinitesimais teremos
\begin{equation}
T(\delta x_{j})|x_{j}\rangle := |x_{j}+\delta x_{j}\rangle.
\end{equation}
Obter $T$ explicitamente será importante para derivarmos as relações de comutação entre posição e momento linear. Como motivação, comecemos considerando o operador de evolução temporal para $H$ independente do tempo e "expandido em primeira ordem" em $t$:
\begin{equation}
U_{\delta t} = e^{-iH\delta t/\hbar} \approx \mathbb{I} - \frac{i}{\hbar}H\delta t.
\end{equation}
Agora notemos que momento linear (velocidade) está relacionado com mudanças de posição e que a unidade de $Ht$ e de $P_{j}x_{j}$ é a mesma para escrever o operador de translação espacial do estado quântico por uma quantidade infinitesimal $dx_{j}$ como 
\begin{equation}
T(\delta x_{j}) = \mathbb{I} - \frac{i}{\hbar}P_{j}\delta x_{j},
\end{equation}
em que $P_{j}$ é o operador Hermitiano que representa a componente $j$ do momento linear.

Considerando-se $\mathcal{O}(\delta x_{j}^{2})\approx 0$, pode-se verificar que o operador de translação espacial $T$, definido na última equação, corresponde ao operador identidade quanto tomamos o limite do deslocamente espacial indo pra zero,
\begin{equation}
\lim_{\delta x_{j}\rightarrow 0}T(\delta x_{j}) = \mathbb{I},
\end{equation}
que este operador obedece uma lei de composicão, para descolamento em uma certa direção,
\begin{equation}
T(\delta x_{j})T(\delta x_{j}') = T(\delta x_{j}+\delta x_{j}'),
\end{equation}
que o seu adjunto é um operador de deslocamento infinitesimal no sentido contrário, 
\begin{equation}
T^{\dagger}(\delta x_{j}) = T(-\delta x_{j}),
\end{equation}
e que é um operador unitário,
\begin{equation}
T^{\dagger}(\delta x_{j})T(\delta x_{j}) = T(\delta x_{j})T^{\dagger}(\delta x_{j}) = \mathbb{I}.
\end{equation}

## Relação de comutação posição-momento linear
Nas próximas demonstrações, a ideia é obter duas expressões para a mesma quantidade. Vamos começar verificando que
\begin{align}
[X_{j},T(\delta x_{j})]|x_{j}\rangle & = X_{j}T(\delta x_{j})|x_{j}\rangle - T(\delta x_{j})X_{j}|x_{j}\rangle \\ 
& = X_{j}|x_{j}+\delta x_{j}\rangle - T(\delta x_{j})x_{j}|x_{j}\rangle \\
& = (x_{j}+\delta x_{j})|x_{j}+\delta x_{j}\rangle - x_{j}|x_{j}+\delta x_{j}\rangle = \delta x_{j}|x_{j}+\delta x_{j}\rangle \\
& = \delta x_{j}T(\delta x_{j})|x_{j}\rangle = \delta x_{j}\left(\mathbb{I} - \frac{i}{\hbar}P_{j}\delta x_{j}\right)|x_{j}\rangle \\
&= \delta x_{j}|x_{j}\rangle - \frac{i}{\hbar}P_{j}(\delta x_{j})^{2}|x_{j}\rangle \\
                               & \approx \delta x_{j}\mathbb{I}|x_{j}\rangle.
\end{align}

Como esta relação vale para qualquer estado da base $\{|x_{j}\rangle\}$, vale para qualquer estado e teremos que $[X_{j},T(\delta x_{j})]=dx_{j}\mathbb{I}$. Mas também podemos escrever
\begin{align}
[X_{j},T(\delta x_{j})] & = [X_{j},\mathbb{I} - \frac{i}{\hbar}P_{j}\delta x_{j}] = [X_{j},\mathbb{I}]  - \frac{i}{\hbar}\delta x_{j} [X_{j},P_{j}] \\
                  & = - \frac{i}{\hbar}\delta x_{j} [X_{j},P_{j}].
\end{align}
Chegamos assim na seguinte relação de comutação posição-momento linear para as componentes:
\begin{equation}
[X_{j},P_{j}] = i\hbar\mathbb{I}.
\end{equation}
Lembrando, $j=1,2,3$.

### Relação de incerteza de Heisenberg
Como vimos anteriormente, um caso particular da relação de incerteza de Heisenberg-Robertson-Schrödinger é
\begin{equation}
\sigma_{A}\sigma_{B} \ge \frac{1}{2}|\langle[A,B]\rangle|,
\end{equation}
que é a relação de incerteza da Robertson. Substituindo, nessa relação, $A$ e $B$ por posição e momento linear, respectivamente, obtemos a (original) _relação de incerteza de Heisenberg_:
\begin{equation}
\sigma_{X_{j}}\sigma_{P_{j}} \ge \frac{\hbar}{2}
\end{equation}
pois
\begin{equation}
|\langle[X_{j},P_{j}]\rangle| = |\langle i\hbar\mathbb{I}\rangle| = \hbar.
\end{equation}

## Translações finitas
Para descrever deslocamentos por uma quantidade finita $\Delta x_{j}$, consideremos o operador de translação espacial associado $T(\Delta x_{j})$ como sendo obtido a partir da composição de um número muito grande $N$ de translações infinitesimais $T(\delta x_{j})$, i.e., $\Delta x_{j}=N\delta x_{j}$ e
\begin{align}
T(\Delta x_{j}) & = T(\delta x_{j})T(\delta x_{j})\cdots T(\delta x_{j}) = \lim_{N\rightarrow\infty}\left(T(\delta x_{j})\right)^{N} \\
                & = \lim_{N\rightarrow\infty}\left(\mathbb{I} - \frac{i}{\hbar}P_{j}\delta x_{j}\right)^{N} = \lim_{N\rightarrow\infty}\left(\mathbb{I} - \frac{iP_{j}\Delta x_{j}/\hbar}{N}\right)^{N}.
\end{align}
Agora, usando $1/n=x/N$ teremos que
\begin{equation}
\lim_{N\rightarrow\infty}\left(1+\frac{x}{N}\right)^{N} =\lim_{n\rightarrow\infty}\left(1+\frac{1}{n}\right)^{nx} =\left(\lim_{n\rightarrow\infty}\left(1+\frac{1}{n}\right)^{n}\right)^{x} = e^{x}.
\end{equation}
Assim,
\begin{equation}
T(\Delta x_{j}) = e^{-iP_{j}\Delta x_{j}/\hbar}.
\end{equation}

## Relações de comutação canônicas
Pense na medida das coordenadas espaciais de um quanton através de um conjunto de pares laser-detector colocados em lados opostos em uma mesma coordenada. É claro que nesse esquema podemos medir duas coordenadas de posição simultâneamente. Por isso, concluímos que as _coordenadas de posição são observáveis compatíveis_, i.e.,
\begin{equation}
[X_{j},X_{k}] = \mathbb{0}\space\forall j,k.
\end{equation}

Vamos mostrar que o mesmo segue para as componentes do momento linear. Para isso notemos que pela composição das translações espaciais finitas,
\begin{align}
& T(\Delta x_{1}\hat{e}_{1})T(\Delta x_{2}\hat{e}_{2}) = T(\Delta x_{2}\hat{e}_{2})T(\Delta x_{1}\hat{e}_{1}) \\ 
& = T(\Delta x_{1}\hat{e}_{1}+\Delta x_{2}\hat{e}_{2}),
\end{align}
(que vale para geometria Euclidiana) e assim
\begin{align}
\mathbb{0} & = [T(\Delta x_{1}\hat{e}_{1}),T(\Delta x_{2}\hat{e}_{2})] = \left[e^{-iP_{1}\Delta x_{1}/\hbar}, e^{-iP_{2}\Delta x_{2}/\hbar}\right] \\
           & = \left[\mathbb{I} -iP_{1}\Delta x_{1}/\hbar + (-iP_{1}\Delta x_{1}/\hbar)^{2}/2! + \cdots, \mathbb{I} -iP_{2}\Delta x_{2}/\hbar + (-iP_{2}\Delta x_{2}/\hbar)^{2}/2! + \cdots\right] \\
           & = \alpha[P_{1},P_{2}] + \beta[P_{1},P_{2}^{2}] + \cdots + \xi[P_{1}^{2},P_{2}] + \cdots.
\end{align}
Usando $[AB,C] = A[B,C] + [A,C]B$ e $[A,BC]=B[A,C]+[A,B]C$, vemos que todos os termos não nulos da última equação serão proporcionais a $[P_{1},P_{2}]$. Portanto, a única maneira dessa relação ser igual ao operador nulo é se essas componentes do momento linear comutarem. Fazendo uma análise análoga para as outras combinações de translações, chegaremos na conclusão de que 
\begin{equation}
[P_{j},P_{k}]=\mathbb{O} \text{ para } j,k=1,2,3.
\end{equation}

_OBS:_ Não podemos usar translações infinitesimais para demonstrar este resultado pois e.g. $[T(dx_{1}\hat{e}_{1}),T(dx_{2}\hat{e}_{2})] = -dx_{1}dx_{2}[P_{1},P_{2}]/\hbar$.

Por fim, baseado nos resultados demonstrados até então, vemos que, como $[X_{j},X_{k\ne j}]=[P_{j},P_{k\ne j}]=\mathbb{O}$, podemos descrever os estados de posição,
\begin{equation}
|\vec{x}\rangle=|x_{1}\rangle\otimes|x_{2}\rangle\otimes|x_{3}\rangle,
\end{equation}
e de momento linear, 
\begin{equation}
|\vec{p}\rangle=|p_{1}\rangle\otimes|p_{2}\rangle\otimes|p_{3}\rangle,
\end{equation}
usando vetores normalizados em 
$$\mathcal{H}=\mathcal{H}_1 \otimes\mathcal{H}_2 \otimes\mathcal{H}_3 .$$
Portanto $X_{j}$ e $P_{k\ne j}$ atuam em espaços diferentes, e por conseguinte comutam:
\begin{equation}
[X_{j},P_{k\ne j}]=\mathbb{O}.
\end{equation}

_OBS._ No caso da posição e momento linear teríamos e.g. $[X_{1}\otimes\mathbb{I}_{2}\otimes\mathbb{I}_{3},\mathbb{I}_{1}\otimes P_{2}\otimes\mathbb{I}_{3}]=\mathbb{O}_{123}$.

Resumindo, obtivemos assim as chamadas __relações de comutação canônicas__:
\begin{align}
& [X_{j},X_{k}] = \mathbb{0}, \\
& [P_{j},P_{k}] = \mathbb{0}, \\
& [X_{j},P_{k}] = i\hbar\mathbb{I}\delta_{jk}, \\
\end{align}
com $j,k=1,2,3$.

## 2ª lei de Newton como um limite da MQ
Para um operador energia potencial sendo uma função bem comportada da posição
\begin{equation}
U=U(\{X_{j}\}),
\end{equation}
na representação de Heisenberg, teremos
\begin{equation}
H^{H}=\sum_{j}\frac{(P_{j}^{H})^{2}}{2m} + U(\{X_{j}^{H}\}).
\end{equation}
Consideremos a eq. de Heisenberg:
\begin{align}
i\hbar\partial_{t}X^{H}_{j} & = [X_{j}^{H},H^{H}]  + i\hbar U_{t}^{\dagger}(\partial_{t}X_{j}^{S})U_{t} \\ 
& = [X_{j}^{H},\frac{1}{2m}\sum_{k}(P_{k}^{H})^{2} + U(\{X_{k}^{H}\})] + 0 \\
& = \frac{1}{2m}\sum_{k}[X_{j}^{H},(P_{k}^{H})^{2}] + [X_{j}^{H},U(\{X_{k}^{H}\})] \\
& = \frac{1}{2m}\sum_{k}\left(P_{k}^{H}[X_{j}^{H},P_{k}^{H}]+[X_{j}^{H},P_{k}^{H}]P_{k}^{H}\right) + 0 \\
& = \frac{1}{2m}\sum_{k}\left(P_{k}^{H}i\hbar\delta_{jk}\mathbb{I}+i\hbar\delta_{jk}\mathbb{I}P_{k}^{H}\right) + 0 \\ 
& = \frac{i\hbar P_{j}^{H}}{m}.
\end{align}
Acima usamos o fato que a relação de comutação é a mesma nas representação de Schrödinger e de Heisenberg e que $\mathbb{I}^{H}=\mathbb{I}^{S}$. Assim
\begin{equation}
m\partial_{t}X^{H}_{j} = P_{j}^{H}.
\end{equation}

Ademais, usando o fato de que $[H^{S},U_{t}]=\mathbb{O}$, pois $H^{S}$ é independente do tempo, teremos
\begin{align}
\partial_{t}H^{H} & = \partial_{t}(U_{t}^{\dagger}H^{S}U_{t}) = \partial_{t}(H^{S}U_{t}^{\dagger}U_{t}) = \partial_{t}H^{S} = 0 \\
& = \partial_{t}\left(\sum_{j}(P_{j}^{H})^{2}/2m + U(\{X_{j}^{H}\})\right) \\ 
& = \frac{1}{2m}\sum_{j}2P_{j}^{H}\partial_{t}P_{j}^{H} + \partial_{t}U(\{X_{j}^{H}\}) \\ 
& = \frac{1}{m}\sum_{j}P_{j}^{H}\partial_{t}P_{j}^{H} + \sum_{j}\partial_{t}X_{j}^{H}\partial_{X_{j}^{H}}U(\{X_{j}^{H}\}) \\
& = \frac{1}{m}\sum_{j}P_{j}^{H}\partial_{t}P_{j}^{H} + \sum_{j}\frac{P_{j}^{H}}{m}\partial_{X_{j}^{H}}U(\{X_{j}^{H}\}) \\
& = \frac{1}{m}\sum_{j}P_{j}^{H}\left(\partial_{t}P_{j}^{H} + \partial_{X_{j}^{H}}U(\{X_{j}^{H}\})\right) \\
\end{align}
Por conseguinte
\begin{equation}
\partial_{t}P_{j}^{H} = - \partial_{X_{j}^{H}}U(\{X_{j}^{H}\}) = F_{j}^{H}.
\end{equation}

Juntando esses resultados teremos que
\begin{align}
& m\partial_{tt}X_{j}^{H} = \partial_{t}P_{j}^{H} = - \partial_{X_{j}^{H}}U(\{X_{j}^{H}\}) = F_{j}^{H} \\
& \therefore m\partial_{tt}\vec{X}_{H} = \partial_{t}\vec{P}_{H} = - \nabla_{\vec{X}_{H}}U(\vec{X}_{H}) = \vec{F}_{H}, \\
& \therefore m\partial_{tt}\langle\psi_{0}^{S}|\vec{X}(t)|\psi_{0}^{S}\rangle = \partial_{t}\langle\psi_{0}^{S}|\vec{P}(t)|\psi_{0}^{S}\rangle = \langle\psi_{0}^{S}|(-\nabla_{\vec{X}}U(\vec{X}))|\psi_{0}^{S}\rangle = \langle\psi_{0}^{S}|\vec{F}(t)|\psi_{0}^{S}\rangle, \\
& \therefore m\partial_{tt}\langle\vec{X}(t)\rangle = \partial_{t}\langle\vec{P}(t)\rangle = \langle - \nabla_{\vec{X}}U(\vec{X})\rangle = \langle\vec{F}(t)\rangle,
\end{align}
onde $\langle O(t)\rangle$ é o valor esperado do observável $O$ no instante de tempo $t$ (em qualquer representação). 

Vemos assim que o centro do pacote de ondas do quanton se move sujeito à força média seguindo a 2ª lei de Newton. Isso implica que a Mecânica Clássica é o limite da Mecânica Quântica em que os quantons são localizados (pacotes de onda estreitos), comportando-se assim como partículas clássicas.

## Operador momento linear na base de posição
Para $|\psi\rangle = \int_{-\infty}^{\infty} dx_{j}\psi(x_{j})|x_{j}\rangle$ e $\delta x_{j}$ infinitesimal teremos
\begin{align}
T(\delta x_{j})|\psi\rangle & = \int_{-\infty}^{\infty} dx_{j}\psi(x_{j}) T(\delta x_{j})|x_{j}\rangle \\
& = \int_{-\infty}^{\infty} dx_{j}\psi(x_{j}) |x_{j}+\delta x_{j}\rangle \\
& = \int_{-\infty}^{\infty} dx_{j}'\psi(x_{j}'-\delta x_{j})|x_{j}'\rangle,
\end{align}
onde fizemos a troca de variável $x_{j}'=x_{j}+\delta x_{j}$. Agora expandimos a função de onda em série de Taylor,
\begin{equation}
\psi(x_{j}'-\delta x_{j}) \approx \psi(x_{j}') - \delta x_{j}\partial_{x_{j}'}\psi(x_{j}'),
\end{equation}
para obter
\begin{equation}
T(\delta x_{j})|\psi\rangle = \int_{-\infty}^{\infty} dx_{j}'\left(\psi(x_{j}') - \delta x_{j}\partial_{x_{j}'}\psi(x_{j}')\right)|x_{j}'\rangle.
\end{equation}
Tomando o produto interno de $|x_{j}\rangle$ com essa equação,
\begin{align}
\langle x_{j}|T(\delta x_{j})|\psi\rangle & = \int_{-\infty}^{\infty} dx_{j}'\left(\psi(x_{j}') - \delta x_{j}\partial_{x_{j}'}\psi(x_{j}')\right)\langle x_{j}|x_{j}'\rangle \\
& = \int_{-\infty}^{\infty} dx_{j}'\left(\psi(x_{j}') - \delta x_{j}\partial_{x_{j}'}\psi(x_{j}')\right)\delta(x_{j}-x_{j}') \\
& = \psi(x_{j}) - \delta x_{j}\partial_{x_{j}}\psi(x_{j}).
\end{align}

Por outro lado, se usamos $T$ em termos de $P_{j}$ obteremos
\begin{align}
\langle x_{j}|T(\delta x_{j})|\psi\rangle & = \langle x_{j}|\left(\mathbb{I} - \frac{i}{\hbar}P_{j}\delta x_{j}\right)|\psi\rangle \\
& = \langle x_{j}|\psi\rangle - \delta x_{j}\frac{i}{\hbar}\langle x_{j}|P_{j}|\psi\rangle \\
& = \psi(x_{j}) - \delta x_{j}\frac{i}{\hbar}\langle x_{j}|P_{j}|\psi\rangle.
\end{align}

Dessas duas expressões, obtemos a ação do __operador momento linear na base de posição__:
\begin{equation}
\langle x_{j}|P_{j}|\psi\rangle = \frac{\hbar}{i}\partial_{x_{j}}\langle x_{j}|\psi\rangle.
\end{equation}

Por hora, vamos deixar de lado a estrutura produto tensorial e escrever estados de posição como
\begin{equation}
|\vec{x}\rangle = |x_{1}\rangle\otimes|x_{2}\rangle\otimes|x_{3}\rangle =: |x_{1},x_{2},x_{3}\rangle 
\end{equation}
e o vetor operador momento linear como
\begin{align}
\vec{P} & = \hat{e}_{1}P_{1}\otimes\mathbb{I}_{2}\otimes\mathbb{I}_{3}+\hat{e}_{2}\mathbb{I}_{1}\otimes P_{2}\otimes\mathbb{I}_{3} + \hat{e}_{3}\mathbb{I}_{1}\otimes\mathbb{I}_{2}\otimes P_{3} \\ 
&=: \sum_{j=1}^{3}\hat{e}_{j}P_{j}.
\end{align}
Seguindo,
\begin{align}
\langle \vec{x}|\vec{P}|\psi\rangle & = \langle x_{1},x_{2},x_{3}|\vec{P}|\psi\rangle \\
& = \sum_{j=1}^{3}\hat{e}_{j}\langle x_{1},x_{2},x_{3}|P_{j}|\psi\rangle = \sum_{j=1}^{3}\hat{e}_{j}\frac{\hbar}{i}\partial_{x_{j}}\langle x_{1},x_{2},x_{3}|\psi\rangle \\
& = \frac{\hbar}{i}\nabla_{\vec{x}}\langle\vec{x}|\psi\rangle.
\end{align}

OBS. Para estados produto $|\psi\rangle=|\xi\rangle\otimes|\eta\rangle\otimes|\zeta\rangle$ teremos e.g.
\begin{align}
\langle\vec{x}|P_{1}|\psi\rangle & = (\langle x_{1}|\otimes\langle x_{2}|\otimes\langle x_{3}|)(P_{1}\otimes\mathbb{I}_{2}\otimes\mathbb{I}_{2})(|\xi\rangle\otimes|\eta\rangle\otimes|\zeta\rangle) \\
& = \langle x_{1}|P_{1}|\xi\rangle\otimes\langle x_{2}|\eta\rangle\otimes\langle x_{3}|\zeta\rangle \\
& = (-i\hbar\partial_{x_{1}}\langle x_{1}|\xi\rangle)\otimes\langle x_{2}|\eta\rangle\otimes\langle x_{3}|\zeta\rangle \\
& = (-i\hbar\partial_{x_{1}}(\langle x_{1}|\otimes\langle x_{2}|\otimes\langle x_{3}|)(|\xi\rangle\otimes|\eta\rangle\otimes|\zeta\rangle) \\
& =-i\hbar\partial_{x_{1}}\langle\vec{x}|\psi\rangle.
\end{align}
Portanto, para esse tipo de estado, teremos $\langle \vec{x}|\vec{P}|\psi\rangle=-i\hbar\nabla_{\vec{x}}\langle\vec{x}|\psi\rangle$. No entanto, para estados do tipo 
\begin{equation}
|\psi\rangle = \iiint C_{\xi,\eta,\zeta}|\xi\rangle\otimes|\eta\rangle\otimes|\zeta\rangle,
\end{equation}
a verificação não é direta.

## Valores médios de posição e de momento linear
Como vimos, o valor médio de um observável $O$ com espectro contínuo também é calculado por $\langle O\rangle_{\psi} = \langle \psi|O|\psi\rangle$. Para uma componente da posição, usando
\begin{equation}
\mathbb{I}=\int dx_{j}|x_{j}\rangle\langle x_{j}|
\end{equation}
(com a integral sobre $\mathbb{R}$), teremos
\begin{align}
\langle X_{j}\rangle_{\psi} & = \langle\psi|X_{j}|\psi\rangle \\
& = \langle\psi|\mathbb{I}X_{j}\mathbb{I}|\psi\rangle \\
& = \iint dx_{j}dx_{j}'\langle\psi|x_{j}\rangle\langle x_{j}|X_{j}|x_{j}'\rangle\langle x_{j}'|\psi\rangle \\ 
& = \iint dx_{j}dx_{j}'\psi^{*}(x_{j})\langle x_{j}|x_{j}'|x_{j}'\rangle\psi(x_{j}') \\
& = \iint dx_{j}dx_{j}'x_{j}'\psi^{*}(x_{j})\delta(x_{j}-x_{j}')\psi(x_{j}') \\
& = \int dx_{j}|\psi(x_{j})|^{2}x_{j} = \int dx_{j}\psi(x_{j})^{*}x_{j}\psi(x_{j}).
\end{align}

Pelo mesmo procedimento, pode-se verificar que para $n\in\mathbb{N}_{>1}$:
\begin{equation}
\langle X_{j}^{n}\rangle_{\psi} = \int dx_{j}|\psi(x_{j})|^{2}x_{j}^{n} = \int dx_{j}\psi(x_{j})^{*}x_{j}^{n}\psi(x_{j}).
\end{equation}

Para o momento linear
\begin{align}
\langle P_{j}\rangle_{\psi} & = \langle\psi|P_{j}|\psi\rangle = \langle\psi|\mathbb{I}P_{j}|\psi\rangle \\ 
& = \int dx_{j}\langle\psi|x_{j}\rangle\langle x_{j}|P_{j}|\psi\rangle \\
& = \int dx_{j}\psi^{*}(x_{j})\left(\frac{\hbar}{i}\partial_{x_{j}}\right)\langle x_{j}|\psi\rangle \\
& = \int dx_{j}\psi^{*}(x_{j})\left(\frac{\hbar}{i}\partial_{x_{j}}\right)\psi(x_{j}).
\end{align}

No caso de $P_{j}^{2}=P_{j}P_{j}$ fazemos
\begin{align}
\langle P_{j}^{2}\rangle_{\psi} & = \langle\psi|P_{j}P_{j}|\psi\rangle = \langle\psi|\mathbb{I}P_{j}\mathbb{I}P_{j}|\psi\rangle \\
& = \iint dx_{j}dx_{j}'\langle\psi|x_{j}\rangle\langle x_{j}|P_{j}|x_{j}'\rangle\langle x_{j}'|P_{j}|\psi\rangle \\
& = \iint dx_{j}dx_{j}'\psi^{*}(x_{j})\langle x_{j}|P_{j}|x_{j}'\rangle\left(\frac{\hbar}{i}\partial_{x_{j}'}\psi(x_{j}')\right) \\
& = \iint dx_{j}dx_{j}'\psi^{*}(x_{j})\langle x_{j}|P_{j}\left(\left(\frac{\hbar}{i}\partial_{x_{j}'}\psi(x_{j}')\right)|x_{j}'\rangle\right) \\
& = \iint dx_{j}dx_{j}'\psi^{*}(x_{j})\frac{\hbar}{i}\partial_{x_{j}}\left(\langle x_{j}|\left(\frac{\hbar}{i}\partial_{x_{j}'}\psi(x_{j}')\right)|x_{j}'\rangle\right) \\
& = \iint dx_{j}dx_{j}'\psi^{*}(x_{j})\frac{\hbar}{i}\partial_{x_{j}}\left(\left(\frac{\hbar}{i}\partial_{x_{j}'}\psi(x_{j}')\right)\langle x_{j}|x_{j}'\rangle\right) \\
& = \int dx_{j}\psi^{*}(x_{j})\frac{\hbar}{i}\partial_{x_{j}}\int dx_{j}'\left(\left(\frac{\hbar}{i}\partial_{x_{j}'}\psi(x_{j}')\right)\delta(x_{j}-x_{j}')\right)  \\
& = \int dx_{j}\psi^{*}(x_{j})\frac{\hbar}{i}\partial_{x_{j}}\left(\frac{\hbar}{i}\partial_{x_{j}}\psi(x_{j})\right) \\
& = \int dx_{j}\psi^{*}(x_{j})\left(\frac{\hbar}{i}\right)^{2}\partial_{x_{j}x_{j}}\psi(x_{j}) \\
& = \int dx_{j}\psi^{*}(x_{j})\left(\frac{\hbar}{i}\partial_{x_{j}}\right)^{2}\psi(x_{j}).
\end{align}
Pelo mesmo procedimento, pode-se verificar que para $n\in\mathbb{N}_{>2}$:
\begin{equation}
\langle P_{j}^{n}\rangle_{\psi} = \int dx_{j}\psi^{*}(x_{j})\left(\frac{\hbar}{i}\partial_{x_{j}}\right)^{n}\psi(x_{j}).
\end{equation}

## Eq. de Schrödinger na base de posição
Considere um Hamiltoniano que inclui a energia cinética e potencial de uma partícula:
\begin{equation}
H_{t} = \frac{P^{2}}{2m} + U_{t},
\end{equation}
em que $m$ é a massa da partícula e $P^{2}=\vec{P}\cdot\vec{P}$. Consideremos agora a equação de Schrödinger para o vetor de estado:
\begin{equation}
i\hbar\partial_{t}|\psi_{t}\rangle = H_{t}|\psi_{t}\rangle.
\end{equation}

Seguindo, tomemos o produto interno de $|\vec{x}\rangle$ com a eq. de Schrödinger:
\begin{equation}
i\hbar\partial_{t}\langle\vec{x}|\psi_{t}\rangle = \langle\vec{x}|\frac{P^{2}}{2m}|\psi_{t}\rangle + \langle\vec{x}|U_{t}|\psi_{t}\rangle.
\end{equation}

Pode-se verificar que 
\begin{equation}
\langle x_{j}|P_{j}^{2}|\psi\rangle = -\hbar^{2}\partial_{x_{j}x_{j}}\psi(x_{j}) \therefore \langle\vec{x}|P_{j}^{2}|\psi\rangle=-\hbar^{2}\partial_{x_{j}x_{j}}\psi(\vec{x}).
\end{equation}
Assim vem que
\begin{align}
\langle\vec{x}|P^{2}|\psi\rangle & = \langle\vec{x}|\sum_{j}P_{j}^{2}|\psi\rangle \\
& = \sum_{j}\langle\vec{x}|P_{j}^{2}|\psi\rangle \\
& = -\hbar^{2}\sum_{j}\partial_{x_{j}x_{j}}\psi(\vec{x}) \\
& = -\hbar^{2}\nabla^{2}_{\vec{x}}\psi(\vec{x}).
\end{align}

Substituindo na eq. de Schrödinger acima, e assumindo
\begin{equation}
U_{t}|\vec{x}\rangle = u_{t}(\vec{x})|\vec{x}\rangle
\end{equation}
com $u_{t}(\vec{x})$ sendo uma função escalar da posição, obtemos a equação de Schrödinger para os coeficientes $\left(c_{\vec{x}}=\langle\vec{x}|\psi_{t}\rangle=\psi_{t}(\vec{x})\right)$ do estado na base de posição $\left(|\psi_{t}\rangle=\int d^{3}x c_{\vec{x}}|\vec{x}\rangle\right)$:
\begin{equation}
i\hbar\partial_{t}\psi_{t}(\vec{x}) = \frac{-\hbar^{2}}{2m}\nabla^{2}_{\vec{x}}\psi_{t}(\vec{x}) + u_{t}(\vec{x})\psi_{t}(\vec{x}).
\end{equation}
Ou seja, a MQ ondulatória de Schödinger está contida dentro do formalismo que estamos estudando nessa disciplina.

# Base de momento linear
Quase tudo que discutimos para a posição também vale para o momento linear. Em particular, 
\begin{equation}
P_{j}=P_{j}^{\dagger},\ P_{j}|p_{j}\rangle=p_{j}|p_{j}\rangle \text{ com } p_{j}\in\mathbb{R}
\end{equation}
e 
\begin{equation}
P_{j}=\int_{-\infty}^{+\infty}p_{j}|p_{j}\rangle\langle p_{j}|dp_{j} \text{ para } j=1,2,3.
\end{equation}
Também podemos escrever o estado do sistema na base de momento linear
\begin{equation}
|\psi_{t}\rangle = \int d^{3}p \psi_{t}(\vec{p})|\vec{p}\rangle,
\end{equation}
com $\psi_{t}(\vec{p}) = \langle\vec{p}|\psi_{t}\rangle$ sendo a função de onda na base do momento linear. Se conhecemos a "forma" da energia potencial $U_{t}$ também podemos obter uma equação diferencial para esses coeficientes. No entanto, se temos $\psi_{t}(\vec{x})$ e queremos obter $\psi_{t}(\vec{p})$, podemos começar olhando para
\begin{align}
\psi_{t}(\vec{p}) & = \langle \vec{p}|\psi_{t}\rangle \\
& = \langle \vec{p}|\int d^{3}x \psi_{t}(\vec{x})|\vec{x}\rangle \\
& = \int d^{3}x \psi_{t}(\vec{x})\langle \vec{p}|\vec{x}\rangle.
\end{align}
Lembrando, $\langle\vec{x}|\vec{P}|\psi\rangle = -i\hbar\vec{\nabla}_{\vec{x}}\langle\vec{x}|\psi\rangle$. Como essa equação vale para qualquer $|\psi\rangle$, podemos escrever
\begin{align}
\langle\vec{x}|\vec{P}|\vec{p}\rangle &= \langle\vec{x}|\vec{p}|\vec{p}\rangle = \vec{p}\langle\vec{x}|\vec{p}\rangle \\ 
& = -i\hbar\vec{\nabla}_{\vec{x}}\langle\vec{x}|\vec{p}\rangle.
\end{align}

Para $C\in\mathbb{C}$, pode-se verificar que essa equação é satisfeita pela seguinte função
\begin{equation}
\langle\vec{x}|\vec{p}\rangle = Ce^{i\vec{p}\cdot\vec{x}/\hbar}.
\end{equation}
Para obter a constante, usamos a representação integral da função delta de Dirac
\begin{equation}
\delta(x-a) = \frac{1}{2\pi}\int_{-\infty}^{+\infty}dp e^{ip(x-a)}
\end{equation}
aplicada em
\begin{align}
\langle\vec{x}|\vec{x}'\rangle & = (\langle x_{1}|\otimes\langle x_{2}|\otimes\langle x_{3}|)(|x_{1}'\rangle\otimes|x_{2}'\rangle\otimes|x_{3}'\rangle) \\
& = \Pi_{j=1}^{3}\langle x_{j}|x_{j}'\rangle \\
& = \Pi_{j=1}^{3}\delta(x_{j}-x_{j}') \\ 
& = \langle\vec{x}|\mathbb{I}|\vec{x}'\rangle = \int d^{3}p \langle\vec{x}|\vec{p}\rangle\langle\vec{p}|\vec{x}'\rangle \\ 
& = \int d^{3}p Ce^{i\vec{p}\cdot\vec{x}/\hbar}C^{*}e^{-i\vec{p}\cdot\vec{x}'/\hbar} \\
& = |C|^{2}\int d^{3}p e^{i\vec{p}\cdot(\vec{x}-\vec{x}')/\hbar} \\ 
& = |C|^{2}\int d^{3}p e^{i(p_1 (x_1-x_1')+p_2 (x_2-x_2')+p_3 (x_3-x_3'))/\hbar} \\ 
& = |C|^{2}(2\pi)^{3}\hbar^{3}\Pi_{j=1}^{3}\frac{1}{2\pi}\int d(p_{j}/\hbar) e^{i(p_{j}/\hbar)(x_{j}-x_{j}')} \\
&= |C|^{2}(2\pi\hbar)^{3}\Pi_{j=1}^{3}\delta(x_{j}-x_{j}').
\end{align}
Portanto, se usamos $C\in\mathbb{R}$ então $C=1/(2\pi\hbar)^{3/2}$ e
\begin{equation}
\langle\vec{x}|\vec{p}\rangle = \frac{1}{(2\pi\hbar)^{3/2}}e^{i\vec{p}\cdot\vec{x}/\hbar}.
\end{equation}

Ou seja, as funções de onda nas bases de posição e de momento linear estão relacionadas pela __transformada de Fourier__:
\begin{align}
\psi(\vec{p}) & = \frac{1}{(2\pi\hbar)^{3/2}}\int d^{3}x e^{-i\vec{p}\cdot\vec{x}/\hbar}\psi(\vec{x}), \\
\psi(\vec{x}) & = \frac{1}{(2\pi\hbar)^{3/2}}\int d^{3}p e^{i\vec{p}\cdot\vec{x}/\hbar}\psi(\vec{p}).
\end{align}

## Pacote Gaussiano
A função de onda que nos fornece o pacote de ondas Gaussiano em uma dimensão centrado em $x=0$ lê-se:
\begin{equation}
\psi_{G}(x) = \frac{1}{\pi^{1/4}d^{1/2}}e^{ikx-x^{2}/2d^{2}}\ \therefore\ |\psi_{G}(x)|^{2} = \frac{1}{\pi^{1/2}d}e^{-x^{2}/d^{2}},
\end{equation}
com $k,d\in\mathbb{R}$. Teremos assim que a função de onda na base de momento linear é
\begin{align}
\psi(p) & = (2\pi\hbar)^{-1/2}\int_{-\infty}^{+\infty}e^{-ipx/\hbar}\psi_{G}(x)dx \\
& = (2\pi\hbar)^{-1/2}\pi^{-1/4}d^{-1/2}\int_{-\infty}^{+\infty}e^{-\hbar^{-1}ipx}e^{ikx-2^{-1}d^{-2}x^{2}}dx \\
& = (2\hbar)^{-1/2}\pi^{-3/4}d^{-1/2}\int_{-\infty}^{+\infty}e^{-i\hbar^{-1}(p-\hbar k)x-2^{-1}d^{-2}x^{2}}dx \\
& = (2\hbar)^{-1/2}\pi^{-3/4}d^{-1/2}\int_{-\infty}^{+\infty}e^{-\left(2^{-1/2}di\hbar^{-1}(p-\hbar k)+2^{-1/2}d^{-1}x\right)^2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}dx,
\end{align}
em que completamos quadrados para obter a última expressão. Fazendo a seguinte mudança de variável: 
$$y = 2^{-1/2}di\hbar^{-1}(p-\hbar k)+2^{-1/2}d^{-1}x\ \therefore\ dy = 2^{-1/2}d^{-1}dx$$
obtemos
\begin{align}
\psi(p) & = (2\hbar)^{-1/2}\pi^{-3/4}d^{-1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}\int_{-\infty}^{+\infty}e^{-y^{2}}dy2^{1/2}d \\
& = \hbar^{-1/2}\pi^{-3/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}\pi^{1/2} \\
& = \hbar^{-1/2}\pi^{-1/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}.
\end{align}

Lembrando, para $n\in\mathbb{N_{>0}}$ temos $\langle X^{n}\rangle_{\psi} = \int_{-\infty}^{\infty}x^{n}|\psi(x)|^{2}dx$ e $\langle P^{n}\rangle_{\psi} = \int_{-\infty}^{\infty}\psi^{*}(x)(-i\hbar\partial_{x})^{n}\psi(x)dx$. Assim
\begin{align}
\langle X\rangle_{\psi_{G}} & = \int_{-\infty}^{+\infty}x|\psi_{G}(x)|^{2}dx \\
& = \frac{1}{\pi^{1/2}d}\int_{-\infty}^{+\infty}xe^{ikx-x^{2}/2d^{2}}e^{-ikx-x^{2}/2d^{2}}dx \\
& = \frac{1}{\pi^{1/2}d}\int_{-\infty}^{+\infty}xe^{-x^{2}/d^{2}}dx \\
& = 0,
\end{align}
pois a função sendo integrada é ímpar. Para o quadrado da posição usa o truque de Feynman:
\begin{align}
\langle X^{2}\rangle_{\psi_{G}} & = \int_{-\infty}^{+\infty}x^{2}|\psi_{G}(x)|^{2}dx = \frac{1}{\pi^{1/2}d}\int_{-\infty}^{+\infty}x^{2}e^{-x^{2}/d^{2}}dx \\
& = -\frac{1}{\pi^{1/2}d}\partial_{d^{-2}}\int_{-\infty}^{+\infty}e^{-d^{-2}x^{2}}dx = -\frac{1}{\pi^{1/2}d}\partial_{d^{-2}}\sqrt{\pi/d^{-2}} \\
& = -\frac{1}{d}\partial_{d^{-2}}(d^{-2})^{-1/2} = -\frac{1}{d}(-1/2)(d^{-2})^{-3/2} \\
& = \frac{d^{2}}{2}.
\end{align}

Pode-se verificar que para o momento linear
\begin{align}
&\langle P\rangle = \hbar k, \\
&\langle P^{2}\rangle = \hbar^{2}k^{2} + \hbar^{2}d^{-2}2^{-1}.
\end{align}

As variâncias são 
\begin{align}
\mathrm{Var}(X)& =\langle X^{2}\rangle-\langle X\rangle^{2} = 2^{-1}d^{2}, \\
\mathrm{Var}(P)&=\langle P^{2}\rangle-\langle P\rangle^{2} = \hbar^{2}d^{-2}2^{-1}.
\end{align}
Isso nos leva a 
\begin{equation}
\mathrm{Var}(X)\mathrm{Var}(P)=\hbar^{2}/4,
\end{equation}
que é o mínimo possível na relação de incerteza de Heisenberg.

```python
def psi_gauss_x(kx,d,x):
    return ((cos(kx*x)+1j*sin(kx*x))*(exp(-x**2/(2*d**2))))/(pi**(1/4)*sqrt(d))

kx, d = symbols('k_{x} d', real=True, positive=True); x = symbols('x', real=True)
norm = integrate(simplify(psi_gauss_x(kx,d,x)*conjugate(psi_gauss_x(kx,d,x))),(x,-oo,+oo)) # ok!
xm = integrate(x*exp(-x**2/d**2)/(sqrt(pi)*d),(x,-oo,+oo)) # valor médio de x
x2m = integrate(x**2*exp(-x**2/d**2)/(sqrt(pi)*d),(x,-oo,+oo)) # valor médio de x^2
norm, xm, x2m
```

## Evolução do pacote gaussiano para uma partícula livre
Consideremos um quanton com função de onda inicial na base de posição 
$$\Psi(x,0) = \psi_{G}(x) = \frac{1}{\pi^{1/4}d^{1/2}}e^{ikx-x^{2}/2d^{2}}.$$
Vimos que neste caso a função de onda na base de momento linear é
$$\Psi(p,0) = \psi(p) = \hbar^{-1/2}\pi^{-1/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}.$$
Se este quanton está livre, o operador hamiltoniano fica escrito como
$$H = \frac{P^2}{2m}.$$

Como podemos escrever qualquer estado como uma combinação linear dos autovetores do operador momento linear, i.e., $|\psi\rangle=\int_{-\infty}^{\infty}dp \psi(p)|p\rangle$, podemos fazer isso para o estado inicial
$$|\Psi_0\rangle = \int_{-\infty}^{\infty}dp \Psi(p,0)|p\rangle.$$
Assim, o estado evoluído no tempo pode ser obtido como segue
\begin{align}
|\Psi_t\rangle & = U_t|\Psi_0\rangle = e^{-iHt/\hbar}|\Psi_0\rangle \\
& = e^{-iP^2 t/2m\hbar}\int_{-\infty}^{\infty}dp \Psi(p,0)|p\rangle \\
& = \int_{-\infty}^{\infty}dp \Psi(p,0)e^{-iP^2 t/2m\hbar}|p\rangle \\
& = \int_{-\infty}^{\infty}dp \Psi(p,0)e^{-ip^2 t/2m\hbar}|p\rangle \\
& = \int_{-\infty}^{\infty}dp \hbar^{-1/2}\pi^{-1/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}e^{-ip^2 t/2m\hbar}|p\rangle.
\end{align}

Usando 
$$\langle x|p\rangle = \frac{1}{(2\pi\hbar)^{1/2}}e^{ipx/\hbar}$$
teremos que
\begin{align}
& \Psi(x,t) = \langle x|\Psi_t\rangle \\
& = \langle x|\int_{-\infty}^{\infty}dp \hbar^{-1/2}\pi^{-1/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}e^{-ip^2 t/2m\hbar}|p\rangle \\
& = \int_{-\infty}^{\infty}dp \hbar^{-1/2}\pi^{-1/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}e^{-ip^2 t/2m\hbar} \langle x|p\rangle \\
& = \int_{-\infty}^{\infty}dp \hbar^{-1/2}\pi^{-1/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}e^{-ip^2 t/2m\hbar} \frac{1}{(2\pi\hbar)^{1/2}}e^{ipx/\hbar} \\
& = \frac{d^{1/2}}{2^{1/2}\pi^{3/4}\hbar}\int_{-\infty}^{\infty}dp e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}e^{-ip^2 t/2m\hbar}e^{ipx/\hbar} \\
& = \frac{d^{1/2}}{2^{1/2}\pi^{3/4}\hbar}\sqrt{\frac{\pi}{b}}\exp\Big(\frac{-2bk^2 d^2 + c^2 }{4b}\Big) \\
& = \Big(\frac{1}{\pi d^2}\Big)^{1/4}\frac{1}{\sqrt{1 + \frac{it\hbar}{md^2}}}\exp\Big(\frac{-i\hbar k^2 d^2 t/2m + ikd^2x - x^2/2 }{d^2\big(1 + \frac{it\hbar}{md^2}\big)}\Big) \\
\end{align}

em que usamos
\begin{align}
I & = \int_{-\infty}^{\infty}dp e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}e^{-ip^2 t/2m\hbar}e^{ipx/\hbar} \\
& = \int_{-\infty}^{\infty}dp e^{-(p^2-2\hbar kp+\hbar^2 k^2)d^2 /2\hbar^2}e^{-ip^2 t/2m\hbar}e^{ipx/\hbar} \\
& = e^{-k^2 d^2 /2}\int_{-\infty}^{\infty}dp e^{-(d^2/2\hbar^2 +it/2m\hbar)p^2 + (kd^2 /\hbar + ix/\hbar)p} \\
& =: e^{-k^2 d^2 /2}\int_{-\infty}^{\infty}dp e^{-(bp^2 - cp)} \\
& =: e^{-k^2 d^2 /2}\int_{-\infty}^{\infty}dp e^{-(\sqrt{b}p - c/2\sqrt{b})^2 + c^2 /4b} \\
& =: e^{-k^2 d^2 /2 + c^2 /4b}\int_{-\infty}^{\infty}\frac{dy}{\sqrt{b}} e^{-y^2} \\
& =: e^{-k^2 d^2 /2 + c^2 /4b}\sqrt{\frac{\pi}{b}} \\
& =: \sqrt{\frac{\pi}{b}}\exp\Big(\frac{-4bk^2 d^2 /2 + c^2 }{4b}\Big) \\
\end{align}
onde definimos
$$b := d^2/2\hbar^2 +it/2m\hbar\ \text{e}\ c:= kd^2 /\hbar + ix/\hbar$$
e fizemos a mudança de variável
$$y = \sqrt{b}p - c/2\sqrt{b}\ \therefore\ dy = \sqrt{b}dp.$$
