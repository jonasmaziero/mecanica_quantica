# Partícula de spin 1/2 em um campo magnético
Considerando o protótipo clássico de dipolo magnético, a espira de corrente, obtemos a seguinte relação entre momento de dipolo magnético e momento angular (__exercício__)
\begin{equation}
\vec{\mu} = \frac{q}{2m}\vec{L}.
\end{equation}
Para descrever "distribuições mais complicadas de corrente", corrigimos esta expressão usando o fator giromagnético $g$, que deve ser determinado experimentalmente para cada sistema físico (e.g. $g \approx −2.0023$ para o elétron, $g \approx 5.586$ para o próton e $g \approx −3.826$ para o nêutron). Assim,
\begin{equation}
\vec{\mu} = \frac{gq}{2m}\vec{L}.
\end{equation}
Quando nos referimos ao momento angular intrínseco, o spin, usamos em geral o símbolo $\vec{S}$, que por sua vez é escrito em termos de um outro operador como $\vec{S}=(\hbar/2)\vec{\Gamma}$. Juntando tudo segue que:
\begin{equation}
\vec{\mu} = \frac{\hbar gq}{4m}\vec{\Gamma}.
\end{equation}

A energia potencial de interação entre um momento e um campo magnético é dada por:
\begin{equation}
U = -\vec{\mu}\cdot\vec{B} = -\frac{\hbar gqB}{4m}\vec{\Gamma}.\hat{n}_{B}  = -\frac{\hbar\omega}{2}\vec{\Gamma}.\hat{n}_{B} = H,
\end{equation}
onde definimos a _frequência de Larmor_
\begin{equation}
\omega = \frac{g q B}{2m},
\end{equation}
que nos fornece a intensidade da interação entre $\vec{\mu}$ e $\vec{B}$. Por sua vez, a força sobre uma partícula interagindo dessa forma será
\begin{equation}
\vec{F} = -\vec{\nabla}U = \sum_{j=x,y,z}\mu_{j}\vec{\nabla}B_{j}.
\end{equation}

## Estados e observáveis de um spin 1/2
Em experimentos de Stern e Gerlach (ESG), observa-se a quantização do spin fazendo-se partículas de spin 1/2 atravessar uma região com um gradiente de campo magnético, e observando que somente dois feixes de partículas bem separados resultavam dessa interação. 

Vamos começar olhando para um detalhe que normalmente não é discutido profundamente em textos de MQ. Nesses textos, quando se explica os ESG considera-se e.g. o gradiente de campo magnético $\vec{B}= \beta z\hat{k}$ com $\beta$ constante. No entanto, esse campo não satisfaz a equação de Maxwell $\nabla\cdot\vec{B}=0$ (__Exercício:__ verifique) e não é portanto fisicamente aceitável. Um exemplo de campo que satisfaz essa eq. é $\vec{B}=\beta(x\hat{i}-z\hat{k})$. Mas é notável que nesse caso a força fica escrita como $\vec{F}=-\nabla U = -\nabla (-\mu\cdot\vec{B}) = \nabla\beta(\mu_{x}x-\mu_{z}z) = \beta(\mu_{x}\hat{i} - \mu_{z}\hat{k})$. Mas essa força vai separar o feixe de partículas em quatro feixes, um para cada possível par de valores das componentes do momento magnético $(\mu_{x},\mu_{z})$. Uma maneira de focar na medida de uma única componente é aplicar um campo constante muito forte na direção na qual queremos medir o momento magnético. Um exemplo de tal campo seria $\vec{B}=\beta x\hat{i}+(\alpha-\beta z)\hat{k}$ com $\beta \ll \alpha$. Observo no entanto que na aplicação na interpretação de medidas em MQ, essa exclusão pela média nula de $F_{x}$ complica a interpretação quântica de colapso, etc.

Consideremos o último exemplo de campo magnético, que nos fornecerá efetivamente $F_{x}=F_{y}=0$ e 
\begin{equation}
F_{z}=\mu_{z}B,
\end{equation}
onde $B := \beta$. Como as partículas são desviadas pra cima ou pra baixo, temos duas possibilidades para os estados preparados medindo o observável "componente de spin na direção $z$", que denotaremos por $\sigma_{z}$. Isso nos indica que
\begin{equation}
\text{Espaço de estados } = \mathbb{C}^{2}.
\end{equation}
Vamos denotar os estados preparados pela medida de $\sigma_{z}$ (que deve ser representado por um operador Hermitiano) por
\begin{equation}
\begin{cases}
\text{ partícula desviada pra cima em z} \Rightarrow |z_{+}\rangle \text{ foi preparado}, \\
\text{ partícula desviada pra baixo em z} \Rightarrow |z_{-}\rangle \text{ foi preparado}. 
\end{cases}
\end{equation}
Normalizaremos os autovalores para obter
\begin{equation}
\begin{cases}
\text{ partícula desviada pra cima em z} \Rightarrow z_{+} = +1, \\
\text{ partícula desviada pra baixo em z} \Rightarrow z_{-} = -1. 
\end{cases}
\end{equation}
Por conseguinte, o operador $\sigma_{z}$ e sua representação matricial (__exercício__) ficam:
\begin{equation}
\sigma_{z} = z_{+}|z_{+}\rangle\langle z_{+}| + z_{-}|z_{-}\rangle\langle z_{-}| = 
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix},
\end{equation}
na base ortonormal $\{|z_{+}\rangle,|z_{-}\rangle\}$.

_OBS:_ Em qualquer direção que se coloque o gradiente de campo magnético, teremos dois feixes desviados com proporções iguais. Por conseguinte, os autovalores terão a mesma magnitude para medida da camponente do spin em qualquer direção.

Determina-se experimentalmente que se prepararmos as partículas no estado $|z_{+}\rangle$ ou $|z_{-}\rangle$ através da medida de $\sigma_{z}$ e depois as fizermos passar por um gradiente de $\vec{B}$ na direção $x$, metade destas serão desviadas pra cima e metade pra baixo. Seremos consistentes com a regra de Born se usarmos
\begin{equation}
|z_{\pm}\rangle = \sqrt{1/2}|x_{+}\rangle \pm \sqrt{1/2}|x_{-}\rangle. \\
\end{equation}

__Exercício:__ Verifique que $|x_{\pm}\rangle = \sqrt{1/2}|z_{+}\rangle \pm \sqrt{1/2}|z_{-}\rangle$ e que eles formam uma base ortonormal.

Nesse caso o observável fica escrito como (__exercício__)
\begin{equation}
\sigma_{x} = x_{+}|x_{+}\rangle\langle x_{+}| + x_{-}|x_{-}\rangle\langle x_{-}| = |z_{+}\rangle\langle z_{-}| + |z_{-}\rangle\langle z_{+}| =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix},
\end{equation}
na base $\{|z_{+}\rangle,|z_{-}\rangle\}$.

Também é um fato empírico que se preparamos as partículas em qualquer um dos estados $|z_{\pm}\rangle,|x_{\pm}\rangle$ e as fizermos passar por um gradiente de $\vec{B}$ na direção $y$, a probabilidade de desvio pra cima ou pra baixo é $1/2$. A única opção que temos agora é usar coeficientes complexos para manter consistência com a regra de Born:
\begin{equation}
|y_{\pm}\rangle = \sqrt{1/2}|z_{+}\rangle \pm i\sqrt{1/2}|z_{-}\rangle,
\end{equation}
com $i=\sqrt{-1}$. Isso nos retorna (__exercício__)
\begin{equation}
\sigma_{y} = y_{+}|y_{+}\rangle\langle y_{+}| + y_{-}|y_{-}\rangle\langle y_{-}| = -i|z_{+}\rangle\langle z_{-}| + i|z_{-}\rangle\langle z_{+}| =
\begin{bmatrix}
0 & -i \\
i & 0
\end{bmatrix},
\end{equation}
na base $\{|z_{+}\rangle,|z_{-}\rangle\}$.

__Exercício:__ Verifique que $|z_{+}\rangle = (1/\sqrt{2})(|y_{+}\rangle+|y_{-}\rangle)$, $|z_{-}\rangle = (-i/\sqrt{2})(|y_{+}\rangle-|y_{-}\rangle)$, $|y_{\pm}\rangle = ((1\pm i)/2)|x_{+}\rangle + ((1\mp i)/2)|x_{-}\rangle$, $|x_{\pm}\rangle = ((1\mp i)/2)|y_{+}\rangle + ((1\pm i)/2)|y_{-}\rangle$.

Agora, para descrever a componente do spin 1/2 em uma _direção qualquer_ consideramos o vetor de Pauli
\begin{equation}
\vec{\sigma}=\hat{i}\sigma_{x}+\hat{j}\sigma_{y}+\hat{k}\sigma_{z}
\end{equation}
e o vetor unitário naquela direção ($\theta\in[0,\pi/2]$, $\phi\in[0,2\pi]$),
\begin{equation}
\hat{n}=\hat{i}n_{x}+\hat{j}n_{y}+\hat{k}n_{z} = \hat{i}\sin\theta\cos\phi+\hat{j}\sin\theta\sin\phi+\hat{k}\cos\theta,
\end{equation}
para escrever (__exercício__)
\begin{equation}
\hat{n}\cdot\vec{\sigma} = \sigma_{x}\sin\theta\cos\phi + \sigma_{y}\sin\theta\sin\phi + \sigma_{z}\cos\theta =
\begin{bmatrix}
\cos\theta & e^{-i\phi}\sin\theta \\
e^{i\phi}\sin\theta & -\cos\theta
\end{bmatrix},
\end{equation}
cujos autovalores são $n_{\pm}=\pm 1$ e autovetores podem ser expressos como (__exercício__):
\begin{align}
|n_{+}\rangle & = \cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle, \\
|n_{-}\rangle & = \sin(\theta/2)|z_{+}\rangle - e^{i\phi}\cos(\theta/2)|z_{-}\rangle.
\end{align}

```python
def Pauli(j):
    if j == 0:
        return Matrix([[1,0],[0,1]])
    elif j == 1:
        return Matrix([[0,1],[1,0]])
    elif j == 2:
        return Matrix([[0,-1j],[1j,0]])
    elif j == 3:
        return Matrix([[1,0],[0,-1]])
th, ph = symbols('theta phi')
NS = sin(th)*cos(ph)*Pauli(1) + sin(th)*sin(ph)*Pauli(2) + cos(th)*Pauli(3)
simplify(NS)
```

```python
NS.eigenvects()
```

### Probabilidades e valores médios para o estado geral $|n_{+}\rangle$
Vamos considerar então que o sistema está preparado no estado $|n_{+}\rangle$. Pela regra de Born, teremos então que para medidas de $\sigma_{z}$,
\begin{align}
Pr(z_{+}|n_{+}) & = |\langle z_{+}|n_{+}\rangle|^{2} \\ 
                & = |\langle z_{+}|(\cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle)|^{2} \\
                & = |\cos(\theta/2)\langle z_{+}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)\langle z_{+}|z_{-}\rangle|^{2} \\
                & = |\cos(\theta/2)(1) + e^{i\phi}\sin(\theta/2)(0)|^{2} \\
                & = \cos^{2}(\theta/2) = (1 +\cos\theta)/2 = (1 + n_{z})/2.\\
Pr(z_{-}|n_{+}) & = |\langle z_{-}|n_{+}\rangle|^{2} = |e^{i\phi}\sin(\theta/2)|^{2} = \sin^{2}(\theta/2) = (1 -\cos\theta)/2 = (1 - n_{z})/2.
\end{align}
Note que poderíamos ter obtido o último resultado usando $Pr(z_{+}|n_{+})+Pr(z_{-}|n_{+})=1$.

Para medidas de $\sigma_{x}$,
\begin{align}
Pr(x_{+}|n_{+}) & = |\langle x_{+}|(\cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle)|^{2} \\
                & = |\cos(\theta/2)\langle x_{+}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)\langle x_{+}|z_{-}\rangle|^{2} \\
                & = |\cos(\theta/2)(1/\sqrt{2}) + e^{i\phi}\sin(\theta/2)(1/\sqrt{2})|^{2} \\
                & = (1/2)\left(\cos(\theta/2) + e^{i\phi}\sin(\theta/2)\right)\left(\cos(\theta/2) + e^{-i\phi}\sin(\theta/2)\right) \\
                & = \left(\cos^{2}(\theta/2) + e^{-i\phi}\sin(\theta/2)cos(\theta/2) + e^{i\phi}\sin(\theta/2)cos(\theta/2) + \sin^{2}(\theta/2)\right)/2 \\
                & = \left(1 + \sin(\theta/2)cos(\theta/2)(\cos\phi -i\sin\phi +\cos\phi + i\sin\phi)\right)/2 \\
                & = \left(1 + 2sin(\theta/2)\cos(\theta/2)\cos\phi\right)/2 \\
                & = \left(1 + sin\theta\cos\phi\right)/2 \\
                & = (1+n_{x})/2. \\
Pr(x_{-}|n_{+}) & = 1-Pr(x_{+}|n_{+}) = (1-n_{x})/2.
\end{align}

Para medidas de $\sigma_{y}$ (__exercício__),
\begin{align}
Pr(y_{+}|n_{+}) & = (1/2)(1 + sin\theta\sin\phi) = (1+n_{y})/2 \\
Pr(y_{-}|n_{+}) & = 1-Pr(y_{+}|n_{+}) = (1-n_{y})/2.
\end{align}

Com isso podemos calcular os _valores médios:_
\begin{align}
\langle\sigma_{z}\rangle_{n_{+}} & = (+1)Pr(z_{+}|n_{+}) + (-1)Pr(z_{-}|n_{+}) \\
                                 & = (1 + n_{z})/2 - (1 - n_{z})/2 \\
                                 & = n_{z}. \\
\langle\sigma_{x}\rangle_{n_{+}} & = (+1)Pr(x_{+}|n_{+}) + (-1)Pr(x_{-}|n_{+}) \\
                                 & = (1 + n_{x})/2 - (1 - n_{x})/2 \\
                                 & = n_{x}. \\
\langle\sigma_{y}\rangle_{n_{+}} & = (+1)Pr(y_{+}|n_{+}) + (-1)Pr(y_{-}|n_{+}) \\
                                 & = (1 + n_{y})/2 - (1 - n_{y})/2 \\
                                 & = n_{y}.
\end{align}

Já os _desvios padrões_ associados a esses observáveis são obtidos como segue (lembrando $\sigma_{O}^{\psi} = \langle O^{2}\rangle_{\psi} - \langle O\rangle_{\psi}^{2}$):
\begin{align}
\langle\sigma_{z}^{2}\rangle_{n_{+}} & = (+1)^{2}Pr(z_{+}|n_{+}) + (-1)^{2}Pr(z_{-}|n_{+}) \\
                                     & = Pr(z_{+}|n_{+}) + Pr(z_{-}|n_{+}) \\
                                     & = 1 \\
                                     & = \langle\sigma_{x}^{2}\rangle_{n_{+}} = \langle\sigma_{y}^{2}\rangle_{n_{+}}, \text{ (exercício)} \\
(\sigma_{\sigma_{z}}^{\psi})^{2} & = \langle \sigma_{z}^{2}\rangle_{\psi} - \langle \sigma_{z}\rangle_{\psi}^{2} \\
                           & = 1-n_{z}^{2} = 1 - \cos^{2}\theta \\
                           & = \sin^{2}\theta, \\
(\sigma_{\sigma_{x}}^{\psi})^{2} & = 1-n_{x}^{2} = 1 - \sin^{2}\theta\cos^{2}\phi, \\
(\sigma_{\sigma_{y}}^{\psi})^{2} & = 1-n_{y}^{2} = 1 - \sin^{2}\theta\cos^{2}\phi.
\end{align}

### Representação de um estado quântico qualquer de $\mathbb{C}^{2}$ em $\mathbb{R}^{3}$
Qualquer estado de um sistema quântico de dois níveis pode ser escrito como $|\psi(\theta,\phi)\rangle=\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle$ com $\theta\in[0,\pi]$ e $\phi\in[0,2\pi]$. 

## Precessão de spin na representação de Schrödinger (RS)
Vamos considerar a evolução temporal do spin 1/2 quando este está submetido a um campo magnético constante na direção $z$, i.e., $\vec{B}=B\hat{k}$. Nesse caso o Hamiltoniano será
\begin{equation}
H = -\frac{\hbar\omega}{2}\vec{\sigma}\cdot\hat{k} = -\frac{\hbar\omega}{2}\sigma_{z} =
\begin{bmatrix}
-\frac{\hbar\omega}{2} & 0 \\
0 & \frac{\hbar\omega}{2}
\end{bmatrix}
= -\frac{\hbar\omega}{2}|z_{+}\rangle\langle z_{+}| + \frac{\hbar\omega}{2}|z_{-}\rangle\langle z_{-}|.
\end{equation}

Como o Hamiltoniano é independente do tempo, o operador de evolução temporal (que é solução de $i\hbar\partial_{t}U_{t}=H_{t}U_{t}$) pode ser escrito como
\begin{equation}
U_{t} = e^{-iHt/\hbar} = e^{i\hbar\omega\sigma_{z}t/2\hbar} = e^{i\omega\sigma_{z}t/2}.
\end{equation}

Vamos considerar que o sistema foi preparado no estado inicial $|n_{+}\rangle$. O estado evoluído do sistema será:
\begin{align}
|\psi_{t}\rangle & = U_{t}|\psi\rangle = U_{t}|n_{+}\rangle \\
                 & = U_{t}\left(\cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle\right) \\
                 & = \cos(\theta/2)U_{t}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)U_{t}|z_{-}\rangle \\
                 & = \cos(\theta/2)e^{i\omega\sigma_{z}t/2}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)e^{i\omega\sigma_{z}t/2}|z_{-}\rangle \\
                 & = \cos(\theta/2)e^{i\omega t/2}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)e^{-i\omega t/2}|z_{-}\rangle.
\end{align}

__OBS:__ Se multiplicarmos o estado do sistema por uma _fase global_, _não modificamos as probabilidades_ pois
\begin{equation}
Pr(o|e^{i\phi}\psi) = |\langle o|e^{i\phi}|\psi\rangle|^{2} = |e^{i\phi}|^{2}|\langle o|\psi\rangle|^{2} = |\langle o|\psi\rangle|^{2} = Pr(o|\psi).
\end{equation}

Em vistas deste fato, vamos considerar
\begin{align}
|\psi_{t}\rangle & \equiv e^{-i\omega t/2}|\psi_{t}\rangle \\
                 & = \cos(\theta/2)|z_{+}\rangle + e^{i(\phi -\omega t)}\sin(\theta/2)|z_{-}\rangle \\
                 & = \cos(\theta_{t}/2)|z_{+}\rangle + e^{i\phi_{t}}\sin(\theta_{t}/2)|z_{-}\rangle.
\end{align}
Com isso podemos usar todos os resultados obtidos anteriormente, com as substituições
\begin{align}
& \theta\rightarrow\theta_{t}=\theta, \\
& \phi\rightarrow\phi_{t}=\phi-\omega t.
\end{align}
Vem então que (lembre que na RS os observáveis não mudam com $t$):
\begin{align}
Pr(z_{+}|\psi_{t}) & = (1+n_{z}(t))/2 = (1+\cos\theta_{t})/2 = (1+\cos\theta)/2 = Pr(z_{+}|\psi) = \text{ constante}, \\
\langle\sigma_{z}\rangle_{\psi_{t}} & = n_{z}(t) = \cos\theta_{t} = \cos\theta = \langle\sigma_{z}\rangle_{\psi}  = \text{ constante}, \\
Pr(x_{+}|\psi_{t}) & = (1+n_{x}(t))/2 = (1+\sin\theta_{t}\cos\phi_{t})/2 = (1+\sin\theta\cos(\phi-\omega t))/2, \\
\langle\sigma_{x}\rangle_{\psi_{t}} & = n_{x}(t) = \sin\theta_{t}\cos\phi_{t} = \sin\theta\cos(\phi-\omega t), \\
Pr(y_{+}|\psi_{t}) & = (1+n_{y}(t))/2 = (1+\sin\theta_{t}\sin\phi_{t})/2 = (1+\sin\theta\sin(\phi-\omega t))/2, \\
\langle\sigma_{y}\rangle_{\psi_{t}} & = n_{y}(t) = \sin\theta_{t}\sin\phi_{t} = \sin\theta\sin(\phi-\omega t).
\end{align}

__Exercício:__ Faça esses mesmos cálculos (estado evoluído, probabilidades, valores médios, desvios padrões) considerando um campo magnético constante aplicado na direção $x$, e o spin 1/2 inicialmente em um estado genérico $|n_{+}\rangle$.

```python
theta, phi, wt, N = symbols('theta phi wt N')
theta = np.pi/2
phi = 0.0
N = 100
dwt = (2.*np.pi-0)/N
xavg = np.zeros(N)
yavg = np.zeros(N)
zavg = np.zeros(N)
wtv = np.zeros(N)
wt = -dwt
for j in range(0,N):
    wt += dwt
    wtv[j] = wt
    xavg[j] = sin(theta)*cos(phi-wtv[j])
    yavg[j] = sin(theta)*sin(phi-wtv[j])
    zavg[j] = cos(theta)
plt.plot(wtv, xavg, label=r'$\langle\sigma_{x}\rangle$')
plt.plot(wtv, yavg, label=r'$\langle\sigma_{y}\rangle$')
plt.plot(wtv, zavg, label=r'$\langle\sigma_{z}\rangle$')
plt.xlabel(r'$\omega t$')
axes = plt.gca()
axes.set_xlim([0,2*np.pi])
axes.set_ylim([-1.02,1.02])
plt.legend()
plt.show()
```

## Precessão de spin na representação de Heisenberg (RH)
Na RH, se o spin 1/2 foi preparado no estado $|n_{+}\rangle$, usaremos este mesmo vetor para calcular probabilidades em tempos posteriores. Por outro lado, evoluiremos os oberváveis. Para isso novamente usaremos 
\begin{equation}
U_{t} = e^{i\omega\sigma_{z}t/2}= e^{i\omega t/2}|z_{+}\rangle\langle z_{+}| + e^{-i\omega t/2}|z_{-}\rangle\langle z_{-}|.
\end{equation}

Para a __componente do spin na direção $z$__,
\begin{align}
|z_{+}(t)\rangle & = U_{t}^{\dagger}|z_{+}\rangle \\
                 & = (e^{-i\omega t/2}|z_{+}\rangle\langle z_{+}| + e^{i\omega t/2}|z_{-}\rangle\langle z_{-}|)|z_{+}\rangle \\
                 & = e^{-i\omega t/2}|z_{+}\rangle\langle z_{+}|z_{+}\rangle + e^{i\omega t/2}|z_{-}\rangle\langle z_{-}|z_{+}\rangle \\
                 & = e^{-i\omega t/2}|z_{+}\rangle, \\
|z_{-}(t)\rangle & = e^{i\omega t/2}|z_{-}\rangle.
\end{align}
Ou seja, este autovetores só ganham uma fase global. Teremos assim que as probabilidades são constantes para medidas de $\sigma_{z}$:
\begin{align}
Pr(z_{+}(t)|n_{+}) & = |\langle z_{+}(t)|n_{+}\rangle|^{2} \\
                   & = |e^{i\omega t/2}\langle z_{+}|n_{+}\rangle|^{2} \\
                   & = |\langle z_{+}|n_{+}\rangle|^{2} \\
                   & = Pr(z_{+}|n_{+}) \\
                   & = (1+n_{z})/2, \\ 
Pr(z_{-}(t)|n_{+}) & = (1-n_{z})/2.
\end{align}
O valor médio e desvio padrão são dados como na RS.

No caso da __componente do spin na direção $x$__ vamos direto para o cálculo do valor médio
\begin{equation}
\langle\sigma_{x}(t)\rangle_{n_{+}} = \langle n_{+}| \sigma_{x}(t)|n_{+}\rangle,
\end{equation}
onde 
\begin{equation}
\sigma_{x}(t) = U_{t}^{\dagger} \sigma_{x} U_{t} = e^{-i\omega\sigma_{z}t/2} \sigma_{x} e^{i\omega\sigma_{z}t/2}.
\end{equation}
Definindo $c=-i\omega t/2$ e usando a fórmula de Baker-Campbell-Hausdroff obtemos (veja a verificação nas notas de aula de Álgebra Linear 04function.ipynb):
\begin{align}
\sigma_{x}(t) & = e^{c\sigma_{z}}\sigma_{x}e^{-c\sigma_{z}} \\
              & = \sigma_{x}\cos(2ci) + \sigma_{y}\sin(2ci)  \\
              & = \sigma_{x}\cos(2(-i\omega t/2)i) + \sigma_{y}\sin(2(-i\omega t/2)i)  \\
              & = \cos(\omega t)\sigma_{x} + \sin(\omega t)\sigma_{y}.
\end{align}

O valor médio é
\begin{align}
\langle\sigma_{x}(t)\rangle_{n_{+}} & = \cos(\omega t)\langle\sigma_{x}\rangle_{n_{+}} + \sin(\omega t)\langle\sigma_{y}\rangle_{n_{+}} \\
                               & = \cos(\omega t)n_{x} + \sin(\omega t)n_{y} \\
                               & = \cos(\omega t)\sin\theta\cos\phi + \sin(\omega t)\sin\theta\sin\phi \\
                               & = \sin\theta(\cos(\omega t)\cos\phi + \sin(\omega t)\sin\phi) \\
                               & = \sin\theta\cos(\phi-\omega t).
\end{align}

__Exercício:__ Obtenha $\langle\sigma_{y}(t)\rangle_{n_{+}}$ usando a fórmula de Baker-Campbell-Hausdorff para calcular $\sigma_{y}(t)$.

#### Precessão de spin a partir da eq. de Heisenberg
Aqui usaremos a eq. de Heisenberg $i\hbar\partial_{t}O_{t}^{H}=[o_{t}^{H},H_{t}^{H}]+i\hbar\partial_{t}^{H}O_{t}^{S}$. Nesse caso $\partial_{t}^{H}O_{t}^{S} = U_{t}^{\dagger}(\partial_{t}^{S}O_{t}^{S})U_{t}=U_{t}^{\dagger}(\mathbb{O})U_{t}=\mathbb{O}$, pois os observáveis considerados, $\{\sigma_{j}^{S}\}_{j=x,y,z}$, não dependem explicitamente do tempo. Usaremos também a relação de comutação geral $[\sigma_{j}^{H},\sigma_{k}^{H}]=2i\epsilon_{jkl}\sigma_{l}^{H}$ e o Hamiltoniano $H^{H}=-\frac{\hbar\omega}{2}\sigma_{z}^{H}$. Teremos assim que
\begin{align}
i\hbar\partial_{t}\sigma_{x}^{H} & = [\sigma_{x}^{H},H^{H}] =[\sigma_{x}^{H},-\frac{\hbar\omega}{2}\sigma_{z}^{H}] = -\frac{\hbar\omega}{2}2i\epsilon_{132}\sigma_{y}^{H} \\
& \Rightarrow \partial_{t}\sigma_{x}^{H} = \omega\sigma_{y}^{H}. \\
i\hbar\partial_{t}\sigma_{y}^{H} & = [\sigma_{y}^{H},H^{H}] =[\sigma_{y}^{H},-\frac{\hbar\omega}{2}\sigma_{z}^{H}] = -\frac{\hbar\omega}{2}2i\epsilon_{231}\sigma_{y}^{H} \\
& \Rightarrow \partial_{t}\sigma_{y}^{H} = -\omega\sigma_{x}^{H}. \\
i\hbar\partial_{t}\sigma_{z}^{H} & = [\sigma_{z}^{H},H^{H}] =[\sigma_{z}^{H},-\frac{\hbar\omega}{2}\sigma_{z}^{H}] = \mathbb{0} \\
& \Rightarrow \sigma_{z}^{H} = \sigma_{z}^{S}.
\end{align}
Uma possibilidade de solução é
\begin{align}
\sigma_{x}^{H} &= \sigma_{x}^{S}\cos\omega t + \sigma_{y}^{S}\sin\omega t, \\
\sigma_{y}^{H} &= -\sigma_{x}^{S}\sin\omega t + \sigma_{y}^{S}\cos\omega t. \\
& \text{ou} \\
& \begin{bmatrix} \sigma_{x}^{H} \\ \sigma_{y}^{H} \end{bmatrix} = \begin{bmatrix} \cos\omega t & \sin\omega t \\ -\sin\omega t & \cos\omega t \end{bmatrix}\begin{bmatrix} \sigma_{x}^{S} \\ \sigma_{y}^{S} \end{bmatrix}.
\end{align}
Note que a matriz que relaciona os "vetores de observáveis" é a matriz de rotação em $\mathbb{R}^{2}$. Claro, os resultados para os valores médios são os mesmos que aqueles obtidos na representação de Schrödinger.
