# Representações da dinâmica quântica
Assim como em Mecânica Clássica podemos descrever movimento de rotação a partir de pontos de vista diferentes mas equivalentes (e.g. partícula em movimento e referencial parado ou partícula parada e referencial em movimento), também usaremos, quando for conveniente, diferentes representações para a dinâmica quântica. 

# Representação de Schrödinger (RS)
Nessa representação o estado do sistema muda com o tempo mas os observáveis não.

Já vimos essa representação quando introduzimos os postulados da MQ. Na RS estados evoluem unitariamente ($U_{t}^{\dagger}U_{t}=U_{t}U_{t}^{\dagger}=\mathbb{I}$):
\begin{equation}
|\psi_{t}^{S}\rangle = U_t |\psi_{0}^{S}\rangle,
\end{equation}
onde $|\psi_{0}^{S}\rangle$ é o estado inicial. O operador que descreve um certo observável não muda com o tempo:
\begin{equation}
O_{t}^{S} = O_{0}^{S} = \sum_{j}o_{j}|o_{j}^{S}\rangle\langle o_{j}^{S}|.
\end{equation}

Se o sistema foi preparado no estado $|\psi_{0}^{S}\rangle$, deixado evoluir segundo seu Hamiltoniano por um tempo $t$, e depois medimos o observável $O$, as probabilidades dos seus possíveis valores são dadas por:
\begin{equation}
Pr(o_{j}(t)) = \left|\langle o_{j}^{S}|\psi_{t}^{S}\rangle\right|^{2}.
\end{equation}

# Representação de Heisenberg (RH)
Nessa representação os observáveis é que mudam com o tempo, e os estados não.

O _objeto essencial_ que liga o formalismo da MQ com o experimento é o conjunto de _probabilidades_ dadas pela regra de Born. Então, desde que não mudemos esses números, podemos mudar a interpretação dos objetos que aparecem no seu cálculo. Por exemplo, na equação acima se em lugar de usar $|\psi_{t}^{S}\rangle=U_{t}|\psi_{0}^{S}\rangle$ usarmos $|\psi_{0}^{S}\rangle$ e no lugar de $|o_{j}^{S}\rangle$ usarmos $U_{t}^{\dagger}|o_{j}^{S}\rangle$, obteremos os mesmos valores para as probabilidades:
\begin{equation}
|(U_{t}^{\dagger}| o_{j}^{S}\rangle)^{\dagger}|\psi_{0}^{S}\rangle|^{2}  = |\langle o_{j}^{S}|(U_{t}|\psi_{0}^{S}\rangle)|^{2} = |\langle o_{j}^{S}|\psi_{t}^{S}\rangle|^{2} = Pr(o_{j}(t)). 
\end{equation}

Portanto, podemos considerar que o estado do sistema é sempre igual àquele preparado em $t=0$ e que são os autovetores do observável que será medido que evoluíram com o tempo, i.e.,
\begin{equation}
|\psi_{t}^{H}\rangle = |\psi_{0}^{S}\rangle
\end{equation}
e
\begin{equation}
|o_{j}^{H}(t)\rangle = U_{t}^{\dagger}|o_{j}^{S}\rangle.
\end{equation}

_OBS._ Tanto na RS quanto na RH o operador de evolução temporal é obtido da equação de Schrödinger $i\hbar\partial_{t}U_{t}=H_{t}U_{t}$.

## Equação de Heisenberg
Esta equação diferencial para a evolução temporal de observáveis é útil e.g. para identificar constantes de movimento (quantidades conservadas) e é obtida como segue. Para começar, substítuimos a equação acima na representação produto externo do observável, obtendo:
\begin{align}
O_{t}^{H} & = \sum_{j}o_{j}|o_{j}^{H}(t)\rangle\langle o_{j}^{H}(t)| = \sum_{j}o_{j}U_{t}^{\dagger}|o_{j}^{S}\rangle\langle o_{j}^{S}|U_{t} = U_{t}^{\dagger}\sum_{j}o_{j}|o_{j}^{S}\rangle\langle o_{j}^{S}|U_{t} \\
     & = U_{t}^{\dagger} O_{0}^{S} U_{t}.
\end{align}
Derivando essa equação em relação ao tempo,
\begin{align}
i\hbar\partial_{t}O_{t}^{H} & = i\hbar\partial_{t}(U_{t}^{\dagger}O_{0}^{S} U_{t}) = -(-i\hbar\partial_{t}U_{t}^{\dagger})O_{0}^{S} U_{t} + 0 + U_{t}^{\dagger}O_{0}^{S}(i\hbar\partial_{t}U_{t}) \\ 
& = -U_{t}^{\dagger} H_{t}O_{0}^{S} U_{t} + U_{t}^{\dagger}O_{0}^{S} H_{t}U_{t} = -U_{t}^{\dagger} H_{t}\mathbb{I}O_{0}^{S} U_{t} + U_{t}^{\dagger}O_{0}^{S}\mathbb{I}H_{t}U_{t} \\ 
& = -U_{t}^{\dagger} H_{t}U_{t}U_{t}^{\dagger}O_{0}^{S} U_{t} + U_{t}^{\dagger}O_{0}^{S}U_{t}U_{t}^{\dagger}H_{t}U_{t} = -H_{t}^{H}O_{t}^{H} + O_{t}^{H}H_{t}^{H} \\
& = [O_{t}^{H},H_{t}^{H}],
\end{align}
que é a _eq. de Heisenberg._ Acima usamos $H_{t}^{\dagger}=H_{t}$ e $-i\hbar\partial_{t}U_{t}^{\dagger}=U_{t}^{\dagger} H_{t}$.

_OBS._ Se $[O_{t}^{H},H_{t}^{H}]=0$ então $\partial_{t}O_{t}^{H}=0$ e $O_{t}^{H}=O_{0}^{S}$ é uma constante de movimento, ou quantidade conservada, i.e., a distribuição de probabilidades, valor médio e desvio padrão associados não mudam durante a evolução temporal do sistema.

### Eq. de Heisenberg para observáveis que dependem explicitamente do tempo na RS
Vamos considerar observáveis que dependam explicitamente do tempo: $O_{t}^{H}=U_{t}^{\dagger}O_{t}^{s}U_{t}$. Teremos assim que
\begin{align}
i\hbar\partial_{t}O_{t}^{H} &= -(-i\hbar\partial_{t}U_{t}^{\dagger})O_{t}^{s}U_{t} + U_{t}^{\dagger}O_{t}^{s}(i\hbar\partial_{t}U_{t}) + i\hbar U_{t}^{\dagger}(\partial_{t}O_{t}^{s})U_{t} \\
&= -U_{t}^{\dagger}H_{t}O_{t}^{s}U_{t} + U_{t}^{\dagger}O_{t}^{s}H_{t}U_{t} + i\hbar U_{t}^{\dagger}(\partial_{t}O_{t}^{s})U_{t} \\
& = [O_{t}^{H},H_{t}^{H}] + i\hbar\partial_{t}^{H}O_{t}^{s},
\end{align}
onde definimos o operador derivada na representação de Heisenberg como
\begin{equation}
\partial_{t}^{H}O_{t}^{s}=U_{t}^{\dagger}(\partial_{t}O_{t}^{s})U_{t}.
\end{equation}
_OBS._ A derivada de um operador é um operador:
\begin{equation}
\partial_{t}O_{t}^{S} := \lim_{\delta t\rightarrow 0}\frac{O_{t+\delta t}^{S}-O_{t}^{S}}{\delta t}.
\end{equation}

_OBS:_ Além do Hamiltoniano, não conheço nenhum exemplo de observável que dependa explicitamente do tempo na RS.

# Representação de interação (RI)
Nessa representação tanto estados quanto os observáveis evoluem com o tempo.

Vamos considerar sistemas físicos para os quais o Hamiltoniano pode ser escrito da seguinte forma:
\begin{equation}
H_{t} = H_{c} + V_{t},
\end{equation}
onde $H_{c} (V_{t})$ é um operador Hermitiano que independe (depende) explicitamente do tempo. 

_OBS._ Veremos que na RI os estados evoluem sob $V_{t}$ e observáveis evoluem sob $H_{c}$.

Usando a parte independente do tempo do Hamiltoniano definiremos o seguinte _operador unitário:_
\begin{equation}
W_{t} = e^{-iH_{c}t/\hbar}.
\end{equation}

Vamos olhar agora para as probabilidades, cujos valores devem independer de representação:
\begin{align}
Pr(o_{j}(t)) & = \left|\langle o_{j}^{S}|\psi_{t}^{S}\rangle\right|^{2} = \left|\langle o_{j}^{S}|\mathbb{I}|\psi_{t}^{S}\rangle\right|^{2} = \left|\langle o_{j}^{S}|W_{t}W^{\dagger}_{t}|\psi_{t}^{S}\rangle\right|^{2} \\
& =: \left|\langle o_{j}^{I}(t)|\psi_{t}^{I}\rangle\right|^{2},
\end{align}
com
\begin{align}
& |\psi_{t}^{I}\rangle = W^{\dagger}_{t}|\psi_{t}^{S}\rangle = W^{\dagger}_{t}U_{t}|\psi_{0}^{S}\rangle \\
& \text{ e } \\
& |o_{j}^{I}(t)\rangle = W^{\dagger}_{t}|o_{j}^{S}\rangle
\end{align}
sendo, respectivamente, definidos como os vetores de estado e autovetores de observáveis evoluídos no tempo na representação de interação.

## Equação de Heisenberg na RI
Os observáveis evoluem na RI como
\begin{align}
O_{t}^{I} & = \sum_{j}o_{j}|o_{j}^{I}(t)\rangle\langle o_{j}^{I}(t)| = \sum_{j}o_{j}W^{\dagger}_{t}|o_{j}^{S}\rangle\langle o_{j}^{S}|W_{t} = W^{\dagger}_{t}\sum_{j}o_{j}|o_{j}^{S}\rangle\langle o_{j}^{S}|W_{t} \\
& = W^{\dagger}_{t} O_{0}^{S} W_{t}.
\end{align}
Derivando essa equação, obtemos
\begin{align}
i\hbar\partial_{t}O_{t}^{I} & = -(-i\hbar\partial_{t}W_{t}^{\dagger}) O^{S} W_{t} + 0 + W^{\dagger}_{t} O^{S} (i\hbar\partial_{t}W_{t}).
\end{align}
Pode-se verificar que $i\hbar\partial_{t}W_{t} = W_{t}H_{c}$. Também usaremos a relação $-i\hbar\partial_{t}W_{t}^{\dagger} = W_{t}^{\dagger}H_{c}$. Com isso obteremos a _eq. de Heisenberg_:
\begin{align}
i\hbar\partial_{t}O_{t}^{I} & = -H_{c}W_{t}^{\dagger} O^{S} W_{t} + W^{\dagger}_{t} O^{S} W_{t}H_{c} = -H_{c}O_{t}^{I} + O_{t}^{I}H_{c} = [O_{t}^{I},H_{c}] \\
& = [O_{t}^{I},H_{c}^{I}],
\end{align}
pois $H_{c}^{I}=H_{c}$ uma vez que $[W_{t},H_{c}]=0$.

## Equação de Schrödinger na RI
Vamos derivar a equação apresentada acima para o estado evoluído na RI em relação ao tempo:
\begin{align}
i\hbar\partial_{t}|\psi_{t}^{I}\rangle & = i\hbar\partial_{t}(W^{\dagger}_{t}U_{t}|\psi_{0}^{S}\rangle) = (i\hbar\partial_{t}W^{\dagger}_{t})U_{t}|\psi_{0}^{S}\rangle + W^{\dagger}_{t}(i\hbar\partial_{t}U_{t})|\psi_{0}^{S}\rangle + 0  \\
& = -H_{c}W^{\dagger}_{t}U_{t}|\psi_{0}^{S}\rangle + W^{\dagger}_{t}H_{t}U_{t}|\psi_{0}^{S}\rangle \\ 
& = -H_{c}W^{\dagger}_{t}U_{t}|\psi_{0}^{S}\rangle + W^{\dagger}_{t}(H_{c}+V_{t})U_{t}|\psi_{0}^{S}\rangle \\ 
& = W^{\dagger}_{t}V_{t}\mathbb{I}U_{t}|\psi_{0}^{S}\rangle = W^{\dagger}_{t}V_{t}W_{t}W^{\dagger}_{t}U_{t}|\psi_{0}^{S}\rangle \\ 
&= V_{t}^{I}|\psi_{t}^{I}\rangle.
\end{align}
Acima usamos $[W^{\dagger}_{t},H_{c}]=0$.

### Conjunto de equações diferenciais para $|\psi_{t}^{I}\rangle$
Vamos considerar que a parte independente do tempo do Hamiltoniano satisfaz a seguinte equação de autovalores e autovetores:
\begin{equation}
H_{c}|E^{c}_{j}\rangle = E^{c}_{j} |E^{c}_{j}\rangle.
\end{equation}
Podemos escrever qualquer vetor de estado do sistema, e em particular $|\psi_{t}^{I}\rangle$, como combinação linear dos autovetores de $H_{c}$:
\begin{equation}
|\psi_{t}^{I}\rangle = \sum_{j=1}^{d} c_{j}(t)|E^{c}_{j}\rangle,
\end{equation}
com $c_{j}(t) := \langle E^{c}_{j}|\psi_{t}^{I}\rangle$. Como os $|E^{c}_{j}\rangle$ não dependem do tempo, vamos tomar o produto interno deles com a equação de Schödinger na RI:
\begin{align}
\langle E^{c}_{j}|i\hbar\partial_{t}|\psi_{t}^{I}\rangle & = \langle E^{c}_{j}|V_{t}^{I}|\psi_{t}^{I}\rangle \\
 i\hbar\partial_{t}\langle E^{c}_{j}|\psi_{t}^{I}\rangle & = \langle E^{c}_{j}|W_{t}^{\dagger}V_{t}W_{t}\sum_{k=1}^{d} c_{k}(t)|E^{c}_{k}\rangle \\
                            i\hbar\partial_{t}c_{j}(t)  & = \sum_{k=1}^{d} c_{k}(t)\langle E^{c}_{j}|e^{iH_{c}t/\hbar}V_{t}e^{-iH_{c}t/\hbar}|E^{c}_{k}\rangle = \sum_{k=1}^{d} c_{k}(t)\langle E^{c}_{j}|e^{iE^{c}_{j}t/\hbar}V_{t}e^{-iE^{c}_{k}t/\hbar}|E^{c}_{k}\rangle \\
                                                        & = \sum_{k=1}^{d} c_{k}(t)e^{i(E^{c}_{j}-E^{c}_{k})t/\hbar}\langle E^{c}_{j}|V_{t}|E^{c}_{k}\rangle = \sum_{k=1}^{d} e^{i\omega_{jk}t}V_{jk}(t) c_{k}(t),
\end{align}
onde definimos as _frequências de Bohr_ como
\begin{equation}
\omega_{jk} = \frac{E^{c}_{j}-E^{c}_{k}}{\hbar}
\end{equation}

e os elementos de matriz da parte dependente do tempo do Hamiltoniano na base de autovetores da parte independente do tempo foram denotados por:
\begin{equation}
V_{jk}(t) = \langle E^{c}_{j}|V_{t}|E^{c}_{k}\rangle.
\end{equation}
Podemos reescrever todas essas $d$ equações em termos de uma equação matricial:
\begin{equation}
i\hbar 
\begin{bmatrix}
\partial_{t}c_{1}(t) \\ \partial_{t}c_{2}(t) \\ \vdots \\ \partial_{t}c_{d}(t)
\end{bmatrix}=
\begin{bmatrix}
V_{11} & V_{12}e^{i\omega_{12}t} & \cdots & V_{1d}e^{i\omega_{1d}t} \\
V_{21}e^{i\omega_{21}t} & V_{22} & \cdots & V_{2d}e^{i\omega_{2d}t} \\
\vdots & \vdots & \ddots & \vdots \\
V_{d1}e^{i\omega_{d1}t} & V_{d2}e^{i\omega_{d2}t} & \cdots & V_{dd}
\end{bmatrix}
\begin{bmatrix}
c_{1}(t) \\ c_{2}(t) \\ \vdots \\ c_{d}(t)
\end{bmatrix}
.
\end{equation}
Acima usamos $\omega_{jj}=0$, $c_{j}(t)=c_{j}$ e $V_{jk}(t) = V_{jk}$. Ademais, note que $\omega_{jk}=-\omega_{kj}$.

Se nosso objetivo for calcular as probabilidades dos autovalores de $H_{c}$, ao resolvermos este sistema de equações lineares teremos resolvido o problema pois
\begin{equation}
Pr(E_{j}^{c}) = |c_{j}(t)|^{2}.
\end{equation}
Por outro lado, se quisermos calcular probabilidades para um observável qualquer, precisamos obter 
\begin{align}
|o_{j}^{I}(t)\rangle &= W_{t}^{\dagger} |o_{j}^{S}\rangle = e^{iH_{c}t/\hbar}|o_{j}^{S}\rangle = e^{iH_{c}t/\hbar}\mathbb{I}|o_{j}^{S}\rangle  = e^{iH_{c}t/\hbar}\sum_{k=1}^{d}|E_{k}^{c}\rangle\langle E_{k}^{c}||o_{j}^{S}\rangle \\ 
& = \sum_{k=1}^{d}e^{iH_{c}t/\hbar}|E_{k}^{c}\rangle\langle E_{k}^{c}|o_{j}^{S}\rangle = \sum_{k=1}^{d}e^{iE_{k}^{c}t/\hbar}\langle E_{k}^{c}|o_{j}^{S}\rangle |E_{k}^{c}\rangle,
\end{align}
para então usá-lo junto com $|\psi_{t}^{I}\rangle$ em
\begin{align}
Pr(o_{j}(t)) & = \left|\langle o_{j}^{I}(t)|\psi_{t}^{I}\rangle\right|^{2} = \left|\sum_{k=1}^{d}e^{-iE_{k}^{c}t/\hbar}\langle E_{k}^{c}|o_{j}^{S}\rangle^{*}\langle E_{k}^{c}|\sum_{j=1}^{d} c_{j}(t)|E^{c}_{j}\rangle\right|^{2} \\
& = \left|\sum_{k=1}^{d}e^{-iE_{k}^{c}t/\hbar}\langle E_{k}^{c}|o_{k}^{S}\rangle^{*}c_{k}(t)\right|^{2}.
\end{align}

# Exercícios

1. Escrevendo o estado evoluído do sistema, na RS, em termos da base de autovetores do observável de interesse,
$|\psi_{t}^{S}\rangle = \sum_{j=1}^{d}c_{j}(t)|o_{j}^{S}\rangle$,
substituindo na equação de Schrödinger $i\hbar\partial_{t}|\psi_{t}^{S}\rangle=H_{t}|\psi_{t}^{S}\rangle$ e tomando o produto interno com $|o_{k}^{S}\rangle$, para $k=1,\cdots,d$, mostre que $i\hbar\partial_{t}c_{k}(t) = \sum_{j=1}^{d}H_{kj}c_{j}(t)$ com $H_{kj}:=\langle o_{k}^{S}|H_{t}|o_{j}^{S}\rangle$. Argumente que essas $d$ equações acopladas podem ser escritas como uma única equação diferencial matricial: $i\hbar\partial_{t}\vec{c}(t)=H_{t}\vec{c}(t)$ identificando $\vec{c}$ e $H_{t}$.

1. Obtenha $|\psi_{t}^{S}\rangle$ do último exercício usando como observável o próprio Hamiltoniano, o considerando independente do tempo, na RS. 

1. Verifique, usando a definição de derivada, que $(i\hbar\partial_{t}U_{t})^{\dagger} = -i\hbar\partial_{t}U_{t}^{\dagger}$. Verifique assim que $-i\hbar\partial_{t}U_{t}^{\dagger}=U_{t}^{\dagger} H_{t}$.

1. Verifique que se o Hamiltoniano de um sistema é independente do tempo, na RS, então $[U_{t},H] = 0$.

1. Consideremos que a relação de comutação entre dois observáveis na representação de Schrödinger é
$[A_{t}^{S},B_{t}^{S}]=C_{t}^{S}$. Mostre que essa relação de comutação é preservada quando mudamos para a representação de Heisenberg, i.e., verifique que $[A_{t}^{H},B_{t}^{H}] = C_{t}^{H}$. Ou seja, se dois observáveis são incompatíveis (compatíveis) na RS, suas versões na RH também serão incompatíveis (compatíveis).

1. Verifique que $W_{t}= e^{-iH_{c}t/\hbar}$ é um operador unitário, com $H_{c}$ sendo um operador Hermitiano.

1. Verifique que $[W_{t},H_{c}]=0$, que $i\hbar\partial_{t}W_{t}=H_{c}W_{t}$ e que $-i\hbar\partial_{t}W_{t}^{\dagger}=W_{t}^{\dagger}H_{c}$.

1. Verifique que se expandirmos o estado evoluído na RI na base de autovetores de $H_{c}$, $|\psi_{t}^{I}\rangle = \sum_{j=1}^{d} c_{j}(t)|E^{c}_{j}\rangle$, podemos escrever: $Pr(E_{j}^{c}(t)) = |c_{j}(t)|^{2}$.
