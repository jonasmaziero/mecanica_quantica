# Prova I

$1.$ Considere um oscilador harmônico quântico (OHQ) preparado no estado coerente:
\begin{equation}
|\alpha\rangle = e^{-|\alpha|^{2}/2}\sum_{n=0}^{\infty}\frac{\alpha^{n}}{\sqrt{n!}}|n\rangle,
\end{equation}
onde $\alpha\in\mathbb{C}$, $N|n\rangle=n|n\rangle$ com $n=0,1,2,\cdots$ e $N=a^{\dagger}a$. Dado que $a|\alpha\rangle=\alpha|\alpha\rangle$, com $a$ sendo o operador de abaixamento do OHQ, verifique que
\begin{equation}
\langle N\rangle_{\alpha} = |\alpha|^{2} \text{ e } Pr(N=n|\alpha) = \frac{\langle N\rangle_{\alpha}^{n}e^{-\langle N\rangle_{\alpha}}}{n!}.
\end{equation}

_Sol._ Para o valor médio
\begin{equation}
\langle N\rangle_{\alpha}=\langle\alpha|N|\alpha\rangle=\langle\alpha|a^{\dagger}a|\alpha\rangle = \langle\alpha|\alpha^{*}\alpha|\alpha\rangle = |\alpha|^{2}\langle\alpha|\alpha\rangle = |\alpha|^{2}.
\end{equation}
Quanto à probabilidade,
\begin{align}
Pr(N=n|\alpha) & = |\langle n|\alpha\rangle|^{2} = \left|\langle n|e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}|m\rangle\right|^{2} = \left|e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}\langle n|m\rangle\right|^{2} = \left|e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}\delta_{n,m}\right|^{2} \\
& = \left|\frac{e^{-|\alpha|^{2}/2}\alpha^{n}}{\sqrt{n!}}\right|^{2} = \frac{e^{-|\alpha|^{2}}|\alpha|^{2n}}{n!} = \frac{e^{-\langle N\rangle_{\alpha}}\langle N\rangle_{\alpha}^{n}}{n!}.
\end{align}

$2.$ Para o OHQ sob a ação do hamiltoniano $H=\hbar\omega(N+\mathbb{I}/2)$ e preparado no estado $|\alpha\rangle$, obtenha seu estado evoluído no tempo e verifique que este é proporcional ao estado coerente $|\alpha e^{-i\omega t}\rangle$.

_Sol._ Usando a decomposição espectral $H=\sum_{n=0}^{\infty}\hbar\omega(n+1/2)|n\rangle\langle n|$, obtemos o operador de evolução temporal ($H$ é independente do tempo): $U_{t}=e^{-iHt/\hbar} = e^{-i\omega t/2}\sum_{n=0}^{\infty}e^{-i\omega tn}|n\rangle\langle n|$. Assim
\begin{align}
|\psi_{t}\rangle & = U_{t}|\psi_{0}\rangle = U_{t}|\alpha\rangle = e^{-i\omega t/2}\sum_{n=0}^{\infty}e^{-i\omega tn}|n\rangle\langle n|e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}|m\rangle = e^{-i\omega t/2}e^{-|\alpha|^{2}/2}\sum_{n,m=0}^{\infty}e^{-i\omega tn}\frac{\alpha^{m}}{\sqrt{m!}}|n\rangle\langle n|m\rangle \\
& = e^{-i\omega t/2}e^{-|\alpha e^{-i\omega t}|^{2}/2}\sum_{n=0}^{\infty}\frac{(\alpha e^{-i\omega t})^{n}}{\sqrt{n!}}|n\rangle = e^{-i\omega t/2}|\alpha e^{-i\omega t}\rangle.
\end{align}

$3.$ Considerando as quadraturas $X=\sqrt{\hbar/2m\omega}(a^{\dagger}+a)$ e $P=i\sqrt{\hbar m\omega/2}(a^{\dagger}-a)$ do OHQ preparado no estado coerente $|\alpha\rangle$, verifique que $Var(X)=\hbar/2m\omega$ e $Var(P)=\hbar m\omega/2$, e que portanto o estado coerente minimiza o limite inferior para o produto das incertezas na relação de incerteza de Heisenberg $Var(X)Var(P)\ge \hbar^{2}/4$. Use $[a,a^{\dagger}]=\mathbb{I}$.

_Sol._ Teremos 
\begin{align}
\langle X\rangle_{\alpha} & = \langle\alpha|\sqrt{\hbar/2m\omega}(a^{\dagger}+a)|\alpha\rangle = \sqrt{\hbar/2m\omega}\langle\alpha|(a^{\dagger}+a)|\alpha\rangle = \sqrt{\hbar/2m\omega}(\alpha^{*}+\alpha), \\
\langle X^{2}\rangle_{\alpha} & = (\hbar/2m\omega)\langle\alpha|(a^{\dagger}+a)^{2}|\alpha\rangle = (\hbar/2m\omega)\langle\alpha|((a^{\dagger})^{2}+a^{\dagger}a+aa^{\dagger}+a^{2})|\alpha\rangle = (\hbar/2m\omega)\langle\alpha|((a^{\dagger})^{2}+a^{\dagger}a+a^{\dagger}a+\mathbb{I}+a^{2})|\alpha\rangle \\ 
& = (\hbar/2m\omega)((\alpha^{*})^{2}+2|\alpha|^{2}+1+\alpha^{2}), \\
Var(X)_{\alpha} & = \langle X^{2}\rangle_{\alpha} - \langle X\rangle_{\alpha}^{2} = (\hbar/2m\omega)[(\alpha^{*})^{2}+2|\alpha|^{2}+1+\alpha^{2} - (\alpha^{*})^{2}-2|\alpha|^{2}-\alpha^{2}] = \hbar/2m\omega. \\
\langle P\rangle_{\alpha} & = \langle\alpha|i\sqrt{\hbar m\omega/2}(a^{\dagger}-a)|\alpha\rangle = i\sqrt{\hbar m\omega/2}\langle\alpha|(a^{\dagger}-a)|\alpha\rangle = i\sqrt{\hbar m\omega/2}(\alpha^{*}-\alpha), \\
\langle P^{2}\rangle_{\alpha} & = -(\hbar m\omega/2)\langle\alpha|(a^{\dagger}-a)^{2}|\alpha\rangle = -(\hbar m\omega/2)\langle\alpha|((a^{\dagger})^{2}-a^{\dagger}a-aa^{\dagger}+a^{2})|\alpha\rangle = -(\hbar m\omega/2)\langle\alpha|((a^{\dagger})^{2}-a^{\dagger}a-a^{\dagger}a-\mathbb{I}+a^{2})|\alpha\rangle \\
& = -(\hbar m\omega/2)((\alpha^{*})^{2}-2|\alpha|^{2}-1+\alpha^{2}), \\
Var(P) & = \langle P^{2}\rangle_{\alpha} - \langle P\rangle_{\alpha}^{2} = -(\hbar m\omega/2)((\alpha^{*})^{2}-2|\alpha|^{2}-1+\alpha^{2} - (\alpha^{*})^{2} + 2|\alpha|^{2} - \alpha^{2}) = \hbar m\omega/2.
\end{align}
Vemos assim que $Var(X)Var(P)=\hbar^{2}/4$.

$4.$ Considerando o OHQ, verifique que $N(a^{\dagger}|n\rangle)=(n+1)(a^{\dagger}|n\rangle)$ e que $a^{\dagger}|n\rangle=\sqrt{n+1}|n+1\rangle$.

_Sol._ Temos que
\begin{align}
N(a^{\dagger}|n\rangle) & = a^{\dagger}aa^{\dagger}|n\rangle = a^{\dagger}(a^{\dagger}a+\mathbb{I})|n\rangle = a^{\dagger}(N+\mathbb{I})|n\rangle \\
& = (n+1)(a^{\dagger}|n\rangle).
\end{align}
Portanto $a^{\dagger}|n\rangle\propto|n+1\rangle =: c|n+1\rangle$. Assim a norma quadrada desse vetor lê-se
\begin{align}
& \langle n|aa^{\dagger}|n\rangle = \langle n|(a^{\dagger}a+\mathbb{I})|n\rangle = (n+1)\langle n|n\rangle \\
& \langle n|c^{*}c|n\rangle = |c|^{2}\langle n|n\rangle.
\end{align}
Considerando $c\in\mathbb{R}$ obtemos, $c=\sqrt{n+1}$ e $a^{\dagger}|n\rangle=\sqrt{n+1}|n+1\rangle$.
