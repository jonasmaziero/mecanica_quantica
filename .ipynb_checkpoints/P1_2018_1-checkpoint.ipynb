# Prova I



$1.$ Verifique que se multiplicarmos o estado de um sistema por uma fase global, i.e., $|\psi\rangle\rightarrow e^{i\phi}|\psi\rangle$, as probabilidades para medidas de qualquer observável não mudam.

_Sol._ Da regra de Born teremos
\begin{align}
\mathrm{Pr}(o|e^{i\phi}\psi) & = |\langle o|e^{i\phi}|\psi\rangle|^{2} \\
                            & = |e^{i\phi}|^{2}|\langle o|\psi\rangle|^{2} \\
                            & = 1|\langle o|\psi\rangle|^{2} \\
                            & = \mathrm{Pr}(o|\psi).
\end{align}

$2.$ Considere uma partícula de spin 1/2 inicialmente preparada no estado $|n_{+}\rangle=\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle$, com $|z_{\pm}\rangle$ sendo os autovetores de $\sigma_{z}$, e submetida a um campo magnético constante na direção $y$, i.e., $H=-2^{-1}\hbar\omega\sigma_{y}$. Calcule o estado evoluído deste sistema e o valor médio e desvio padrão do observável $(\sigma_{x}+\sigma_{z})/\sqrt{2}$ nesse estado. Lembrando, $\sigma_{x}=\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$, $\sigma_{y}=\begin{bmatrix} 0 & -i \\ i & 0 \end{bmatrix}$ e $\sigma_{z}=\begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}$ na base $\{|z_{\pm}\rangle\}$.

_Sol._ O estado evoluído pode ser obtido de
\begin{align}
|\psi_{t}\rangle & = U_{t}|n_{+}\rangle = e^{-iHt/\hbar}|n_{+}\rangle = e^{i\omega\sigma_{y}t/2}|n_{+}\rangle \\
                 & = (e^{i\omega t/2}|y_{+}\rangle\langle y_{+}| + e^{-i\omega t/2}|y_{-}\rangle\langle y_{-}|)|n_{+}\rangle \\
                 & = \langle y_{+}|n_{+}\rangle e^{i\omega t/2}|y_{+}\rangle + \langle y_{-}|n_{+}\rangle e^{-i\omega t/2}|y_{-}\rangle \\
                 & = 2^{-1/2}\left(\langle y_{+}|n_{+}\rangle e^{i\omega t/2} + \langle y_{-}|n_{+}\rangle e^{-i\omega t/2}\right)|z_{+}\rangle + 2^{-1/2}i\left(\langle y_{+}|n_{+}\rangle e^{i\omega t/2} - \langle y_{-}|n_{+}\rangle e^{-i\omega t/2}\right)|z_{-}\rangle.
\end{align}
onde usamos $|y_{\pm}\rangle=(|z_{+}\rangle\pm i|z_{-}\rangle)/\sqrt{2}$. Para $\phi=0$ teremos $\langle y_{\pm}|n_{+}\rangle = 2^{-1/2}\left(\cos(\theta/2)\mp i\sin(\theta/2)\right)$. Assim
\begin{align}
|\psi_{t}\rangle & = 2^{-1/2}2\mathrm{Re}\left(\langle y_{+}|n_{+}\rangle e^{i\omega t/2}\right)|z_{+}\rangle + i2^{-1/2}2i\mathrm{Im}\left(\langle y_{+}|n_{+}\rangle e^{i\omega t/2}\right)|z_{-}\rangle \\
                 & = \left(\cos(\theta/2)\cos(\omega t/2)+\sin(\theta/2)\sin(\omega t/2)\right)|z_{+}\rangle - \left(\cos(\theta/2)\sin(\omega t/2)-\sin(\theta/2)\cos(\omega t/2)\right)|z_{-}\rangle \\
                 & = \cos\frac{\theta-\omega t}{2} |z_{+}\rangle + \sin\frac{\theta-\omega t}{2}|z_{-}\rangle \\
                 & = \cos(\theta_{t}/2) |z_{+}\rangle + \sin(\theta_{t}/2)|z_{-}\rangle,
\end{align}
com $\theta_{t}=\theta-\omega t$.

_Valor médio_ de $O=(\sigma_{x}+\sigma_{z})/\sqrt{2} = \hat{n}\cdot\vec{\sigma}$ com $\hat{n}=(1/\sqrt{2},0,1/\sqrt{2})$, que equivale a $\phi=0$ e $\theta = \pi/4$:
\begin{align}
\langle O_{t}\rangle & = \langle\psi_{t}|O|\psi_{t}\rangle \\
                     & = 2^{-1/2}\langle\psi_{t}|(\sigma_{x}+\sigma_{z})(\cos(\theta_{t}/2)|z_{+}\rangle + \sin(\theta_{t}/2)|z_{-}\rangle) \\  
                     & = 2^{-1/2}\langle\psi_{t}|\left(\cos(\theta_{t}/2)|z_{-}\rangle + \sin(\theta_{t}/2)|z_{+}\rangle + \cos(\theta_{t}/2)|z_{+}\rangle - \sin(\theta_{t}/2)|z_{-}\rangle\right) \\
                     & = 2^{-1/2}[\cos(\theta_{t}/2)\langle z_{+}| + \sin(\theta_{t}/2)\langle z_{-}|][(\cos(\theta_{t}/2) - \sin(\theta_{t}/2))|z_{-}\rangle + (\cos(\theta_{t}/2)+\sin(\theta_{t}/2))|z_{+}\rangle ] \\
                     & = 2^{-1/2}[\cos^{2}(\theta_{t}/2) + 2\sin(\theta_{t}/2)\cos(\theta_{t}/2) - \sin^{2}(\theta_{t}/2) ] \\
                     & = 2^{-1/2}[\cos(\theta_{t}) + \sin(\theta_{t}) ]. \\
\end{align}

_Desvio padrão:_ Notando que $O^{2} = \sigma_{0}$,
\begin{align}
\sigma_{O_{t}} & = \sqrt{\langle O_{t}^{2}\rangle - \langle O_{t}\rangle^{2}} \\
               & = \sqrt{1 - 2^{-1}[\cos\theta_{t} + \sin\theta_{t}]^{2}} \\
               & = \sqrt{1 - 2^{-1}[1 + \sin(2\theta_{t})]}.
\end{align}

$3.$ Demonstre, com detalhes, que $W_{t} = e^{-iH_{c}t/\hbar}$ é um operador unitário, dado que $H_{c}$ é um operador Hermitiano.

_Sol._ Considere a decomposição espectral $H_{c} = \sum_{j}E_{j}^{c}|E_{j}^{c}\rangle\langle E_{j}^{c}|$. Assim $W_{t} = e^{-iH_{c}t/\hbar}=\sum_{j}e^{-iE_{j}^{c}t/\hbar}|E_{j}^{c}\rangle\langle E_{j}^{c}|$ e $W_{t}^{\dagger} = \sum_{j}e^{iE_{j}^{c}t/\hbar}|E_{j}^{c}\rangle\langle E_{j}^{c}|$. Como $H_{c}$ é Hermitiano, temos que $\{|E_{j}^{c}\rangle\}$ é uma base ortonormal. Segue então que
\begin{align}
W_{t}W_{t}^{\dagger} & = \sum_{j}e^{-iE_{j}^{c}t/\hbar}|E_{j}^{c}\rangle\langle E_{j}^{c}|\sum_{k}e^{iE_{k}^{c}t/\hbar}|E_{k}^{c}\rangle\langle E_{k}^{c}| \\
                     & = \sum_{j,k}e^{-i(E_{j}^{c}-E_{k}^{c})t/\hbar}|E_{j}^{c}\rangle\langle E_{j}^{c}|E_{k}^{c}\rangle\langle E_{k}^{c}| \\
                     & = \sum_{j,k}e^{-i(E_{j}^{c}-E_{k}^{c})t/\hbar}|E_{j}^{c}\rangle\delta_{jk}\langle E_{k}^{c}| \\
                     & = \sum_{j}1|E_{j}^{c}\rangle\langle E_{j}^{c}| = \mathbb{I}.
\end{align}

$4.$ Verifique que se o Hamiltoniano é independente do tempo, o operador de evolução temporal pode ser escrito como $U_{t}=e^{-iHt/\hbar}$ e o valor médio e desvio padrão de $H$ são constantes.

_Sol._ Dado que o Hamiltoniano é independente do tempo, usamos a decomposição espectral $H=\sum_{j}E_{j}|E_{j}\rangle\langle E_{j}|$ para escrever $U_{t}=\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}|$. Assim
\begin{align}
i\hbar\partial_{t}U_{t} & = i\hbar\partial_{t}\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                        & = i\hbar\sum_{j}\frac{-iE_{j}}{\hbar}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                        & = \sum_{j}E_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                        & = \sum_{k}E_{k}|E_{k}\rangle\langle E_{k}|\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                        & = HU_{t}.
\end{align}

$5.$ Para a mistura estatística $\rho=\sum_{j}p_{j}|\psi_{j}\rangle\langle\psi_{j}|$, verifque que $Pr(o|\rho) = \mathrm{Tr}(\rho|o\rangle\langle o|)$, com a função traço definida como $\mathrm{Tr}(X)=\sum_{j}\langle b_{j}|X|b_{j}\rangle$ para $\{|b_{j}\rangle\}$ qualquer base ortonormal do espaço de Hilbert sendo considerado.

_Sol._ Usa o fato de que o traço é invariante por mudança de base para escolher uma base ortonormal tal que $|b_{1}\rangle = |o\rangle$. Assim teremos $\mathrm{Tr}(\rho|o\rangle\langle o|)=\sum_{j}\langle b_{j}|\rho|o\rangle\langle o|b_{j}\rangle=\sum_{j}\langle b_{j}|\rho|o\rangle\delta_{1j}=\langle o|\rho|o\rangle = Pr(o|\rho)$.
