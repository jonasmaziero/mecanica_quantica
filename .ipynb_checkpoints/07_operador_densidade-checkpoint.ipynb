# Operador densidade e entropia
Este texto é baseado em [https://arxiv.org/abs/1502.04489].

## O operador densidade
Vimos que o significado de dizermos que um sistema quântico foi preparado em um determinado estado $|\psi\rangle$ é que sabemos que um observável foi medido e que se selecionou o autovalor correspondente a este autovetor. Vamos supor que temos somente informação parcial sobre a preparação do sistema. Ou seja, considere que sabemos que o sistema foi preparado no estado $|\psi_{j}\rangle$ com probabilidade $p_{j}$, mas para cada "partícula" que nos é apresentada não sabemos qual é o valor de $j$ (i.e., o índice $j$ é uma variável aleatória). Como podemos calcular probabilidades de medidas futuras de um observável $O$? Bom, para cada medida que faremos de $O$, o sistema estará em um dos estados $\{|\psi_{j}\rangle\}$. Então, para o conjunto de partículas no estado $|\psi_{j}\rangle$, segue valendo a regra de Born:
\begin{equation}
Pr(o|\psi_{j}) = |\langle\psi_{j}|o\rangle|^{2}.
\end{equation}
Para obter a probabilidade para todo o ensemble $\{|\psi_{j}\rangle,p_{j}\}$ consideramos a média dessas probabilidades:
\begin{align}
Pr(o|\{p_{j},|\psi_{j}\rangle\}) & = \sum_{j} Pr(o|\psi_{j})p_{j} = \sum_{j} \langle o|\psi_{j}\rangle\langle\psi_{j}|o\rangle p_{j} = \langle o|\sum_{j} p_{j}|\psi_{j}\rangle\langle\psi_{j}|o\rangle \\
                                 & =: \langle o|\rho|o\rangle,
\end{align}
onde definimos o chamado _operador densidade_ como
\begin{equation}
\rho := \sum_{j} p_{j}|\psi_{j}\rangle\langle\psi_{j}|.
\end{equation}
_OBS:_ Os estados $\{|\psi_{j}\rangle\}$ não precisam ser mutuamente ortogonais. Ademais, o índice $j=1,\cdots,N$ e $N$ pode ser menor, igual ou maior que a dimensão do espaço de estados do sistema $d=\dim\mathcal{H}$.

_OBS:_ Podem existir vários ensembles $\{|\psi_{j}\rangle,p_{j}\}$ diferentes que nos levam ao mesmo operador densidade. Ou seja, há arbitrariedade quando fazemos o caminho $\rho\rightarrow\{|\psi_{j}\rangle,p_{j}\}$. 

_OBS:_ No contexto do operador densidade, temos dois tipos de incerteza, a "clássica", codificada nas probabilidades dos diferentes estados do ensemble, e a "quântica", devida ao fato de que estes estados, individualmente, podem ser superposições dos autovetores do observável a ser medido.

_OBS:_ Repetibilidade continua sendo utilizada para identificar estados preparados.

__Exercício:__ Verifque que
\begin{equation}
Pr(o|\rho) = \mathrm{Tr}(\rho|o\rangle\langle o|) =: \mathrm{Tr}(\rho\Pi_{o}),
\end{equation}
com a função traço definida como $\mathrm{Tr}(X)=\sum_{j}\langle b_{j}|X|b_{j}\rangle$ para $\{|b_{j}\rangle\}$ qualquer base ortonormal do espaço de Hilbert sendo considerado.

__Exercício:__ Verifique que o valor médio de um observável $O$ (com espectro contínuo ou discreto) para um sistema preparado no estado $\rho$ pode ser escrito como:
\begin{equation}
\langle O\rangle_{\rho} = \mathrm{Tr}(\rho O).
\end{equation}

### Propriedades de $\rho$
* Traço unitário <br>
\begin{align}
\mathrm{Tr}(\rho) & = \sum_{j} p_{j}\mathrm{Tr}(|\psi_{j}\rangle\langle\psi_{j}|) = \sum_{j} p_{j} \\
                  & = 1.
\end{align}

* Hermiticidade/Positividade <br>
Primeiramente notemos que qualquer operador densidade é um operador Hermitiano:
\begin{align}
\rho^{\dagger} & = \left(\sum_{j} p_{j}|\psi_{j}\rangle\langle\psi_{j}|\right)^{\dagger} = \sum_{j} (p_{j})^{*}(|\psi_{j}\rangle\langle\psi_{j}|)^{\dagger} = \sum_{j} p_{j}|\psi_{j}\rangle\langle\psi_{j}| \\
               & = \rho.
\end{align}

Então, podemos diagonalizar $\rho = \sum_{j}r_{j}|r_{j}\rangle\langle r_{j}|$. Claro, podemos interpretar esse $\rho$ como representando o ensemble $\{|r_{j}\rangle,r_{j}\}$. Por isso os autovalores $r_{j}$ devem ser positivos (além de $\sum_{j}r_{j}=1$). Garantimos positividade dos autovalores exigindo de forma geral que
\begin{equation}
\rho\ge \mathbb{O}.
\end{equation}



### Exemplo
Para partículas de spin 1/2, consideremos que alguém prepara $|z_{+}\rangle$ com probabilodade $p$ ou $|x_{+}\rangle$ com probabilodade $1-p$. O operador densidade associado a esse ensemble $\{(|z_{+}\rangle,p),(|z_{-}\rangle,1-p)\}$ será:
\begin{align}
\rho & = p|z_{+}\rangle\langle z_{+}| + (1-p)|x_{+}\rangle\langle x_{+}| = p|z_{+}\rangle\langle z_{+}| + 2^{-1}(1-p)(|z_{+}\rangle+|z_{-}\rangle)(\langle z_{+}|+\langle z_{-}|) \\
     & = p|z_{+}\rangle\langle z_{+}| + 2^{-1}(1-p)(|z_{+}\rangle\langle z_{+}|+|z_{-}\rangle\langle z_{+}| + |z_{+}\rangle\langle z_{-}|+|z_{-}\rangle\langle z_{-}|) \\
     & = 2^{-1}(1+p)|z_{+}\rangle\langle z_{+}| + 2^{-1}(1-p)|z_{-}\rangle\langle z_{-}| + 2^{-1}(1-p)(|z_{-}\rangle\langle z_{+}| + |z_{+}\rangle\langle z_{-}|) \\
     & = \frac{1}{2}\begin{bmatrix} 1+p & 1-p \\ 1-p & 1-p \end{bmatrix}.
\end{align}
Acima usamos $|x_{\pm}\rangle = (|z_{+}\rangle \pm |z_{-}\rangle)/\sqrt{2}$.

Probabilidades para medidas de $\sigma_{y}$:
\begin{align}
Pr(y_{+}|\rho) & = \langle y_{+}|\rho| y_{+}\rangle = p\langle y_{+}|z_{+}\rangle\langle z_{+}|y_{+}\rangle + (1-p)\langle y_{+}|x_{+}\rangle\langle x_{+}|y_{+}\rangle = p|\langle z_{+}|y_{+}\rangle|^{2} + (1-p)|\langle x_{+}|y_{+}\rangle|^{2} \\
               & = p|1/\sqrt{2}|^{2} + (1-p)|(1+i)/2|^{2} = p/2 + (1-p)/2 = 1/2 \\
Pr(y_{-}|\rho) & = 1-Pr(y_{+}|\rho)=1/2.          
\end{align}
Acima usamos $|y_{+}\rangle = (|z_{+}\rangle + i|z_{-}\rangle)/\sqrt{2}$.

__Exercício:__ Considere que $|z_{+}\rangle$ é preparado com probabilidade $3/4$ e que $|y_{+}\rangle$ é preparado como probabilidade $1/4$. Obtenha a matriz densidade, na base padrão, para o operador densidade que representa esse ensemble. Usse esse operador para calcular as probabilidades para a componente de spin na direção $\hat{n} = (1/\sqrt{2},0,1/\sqrt{2})$.

## Equação de von Neumann (EvN)
Considere um sistema preparado no estado $\rho = \sum_{j} p_{j}|\psi_{j}\rangle\langle\psi_{j}| \equiv \{|\psi_{j}\rangle,p_{j}\}$. Digamos que este sistema evolui por um tempo $t$ segundo o Hamiltoniano $H_{t}$. Qual será o operador densidade evoluído desse sistema? Da ideia de ensemble sabemos que é cada um dos estados $|\psi_{j}\rangle$ que evoluirá no tempo. Ou seja, no tempo $t$ teremos o novo ensemble $\{|\psi_{j}(t)\rangle,p_{j}\}$ com
\begin{equation}
|\psi_{j}(t)\rangle = U_{t}|\psi_{j}\rangle.
\end{equation}
Então
\begin{align}
\rho_{t} & = \sum_{j} p_{j}|\psi_{j}(t)\rangle\langle\psi_{j}(t)| = \sum_{j} p_{j}U_{t}|\psi_{j}\rangle\langle\psi_{j}|U_{t}^{\dagger} = U_{t}\sum_{j} p_{j}|\psi_{j}\rangle\langle\psi_{j}|U_{t}^{\dagger} \\
         & = U_{t}\rho U_{t}^{\dagger}.
\end{align}

Derivando essa equação em relação ao tempo e usando a eq. de Schrödinger,
\begin{equation}
i\hbar\partial_{t}U_{t} = H_{t}U_{t},
\end{equation}
obteremos a EvN:
\begin{align}
i\hbar\partial_{t}\rho_{t} & = i\hbar\partial_{t}(U_{t})\rho U_{t}^{\dagger} + 0 + U_{t}\rho i\hbar\partial_{t}(U_{t}^{\dagger}) = H_{t}U_{t}\rho U_{t}^{\dagger} - U_{t}\rho U_{t}^{\dagger}H_{t} = H_{t}\rho_{t} - \rho_{t}H_{t} \\
                           & = [H_{t},\rho_{t}].
\end{align}

### Exemplo
As matrizes $\{\sigma_{j}/\sqrt{2}\}_{j=0}^{3}$ formam uma base ortonormal para $\mathbb{C}^{2x2}$, sob o produto interno de Hilbert-Schmidt $\langle A|B\rangle = \mathrm{Tr}(A^{\dagger}B)$. Por conseguinte, qualquer operador densidade pode ser escrito como
\begin{align}
\rho & = \sum_{j=0}^{3} \mathrm{Tr}\left(\rho\frac{\sigma_{j}}{\sqrt{2}}\right)\frac{\sigma_{j}}{\sqrt{2}} = \frac{1}{2}\sum_{j=0}^{3}\langle\sigma_{j}\rangle_{\rho}\sigma_{j} =: \frac{1}{2}\left(\sigma_{0} + \sum_{j=1}^{3}r_{j}\sigma_{j}\right) \\
     & = \frac{1}{2}\begin{bmatrix} 1+r_{3} & r_{1}-i r_{2} \\ r_{1}-i r_{2} & 1-r_{3}  \end{bmatrix},
\end{align}
em que usamos $\langle\sigma_{0}\rangle_{\rho}=1$. <br>
__Exercício:__ Verifique que os autovalores de $\rho$ são $(1\pm r)/2$, com $r = \sqrt{\sum_{j=1}^{3}r_{j}^{2}}$. <br>
_OBS:_ Pela positividade do operador densidade estados físicos estão dentro de uma bola de raio 1 em $\mathbb{R}^{3}$ centrada em (0,0,0). As componentes do ponto de $\mathbb{R}^{3}$ que representa $\rho$ são $(r_{1},r_{2},r_{3})$, que é chamado de _vetor de Bloch_.

Consideremos um spin 1/2 em um estado qualquer $\rho_{t}=\frac{1}{2}\left(\sigma_{0} + \sum_{j=1}^{3}r_{j}(t)\sigma_{j}\right)$ e evoluindo sob o Hamiltoniano $H=-2^{-1}\hbar\omega\sigma_{3}$. Teremos assim que
\begin{align}
i\hbar\partial_{t}\rho_{t} & = [H_{t},\rho_{t}] \\
i\hbar\partial_{t}\frac{1}{2}\begin{bmatrix} 1+r_{3}(t) & r_{1}(t)-i r_{2}(t) \\ r_{1}(t)-i r_{2}(t) & 1-r_{3}(t)  \end{bmatrix} & = [-2^{-1}\hbar\omega\sigma_{3},2^{-1}(\sigma_{0}+r_{1}(t)\sigma_{1}+r_{2}(t)\sigma_{2}+r_{3}(t)\sigma_{3})] \\
\frac{i\hbar}{2}\begin{bmatrix} 0+\partial_{t}r_{3}(t) & \partial_{t}r_{1}(t)-i \partial_{t}r_{2}(t) \\ \partial_{t}r_{1}(t)-i \partial_{t}r_{2}(t) & 0-\partial_{t}r_{3}(t)  \end{bmatrix} & = -\frac{\hbar\omega}{4}[\sigma_{3},\sigma_{0}+r_{1}(t)\sigma_{1}+r_{2}(t)\sigma_{2}+r_{3}(t)\sigma_{3}] \\
                                   & = -\frac{\hbar\omega}{4}\left(\mathbb{0} + r_{1}(t)2i\epsilon_{312}\sigma_{2}+ r_{2}(t)2i\epsilon_{321}\sigma_{1}+\mathbb{0}\right) \\
                                   & = -\frac{i\hbar\omega}{2}\left(r_{1}(t)\sigma_{2} - r_{2}(t)\sigma_{1}\right) \\
                                   & = \frac{i\hbar\omega}{2}\begin{bmatrix} 0 & r_{2}(t)+ir_{1}(t) \\ r_{2}(t)-ir_{1}(t) & 0 \end{bmatrix} \\
\end{align}
Primeiro observe que $\partial_{t}r_{3}(t) = 0 \Rightarrow r_{3}(t)=r_{3}$.
Resta-nos resolver o seguinte sistema de eqs. diferenciais lineares de primeira ordem acopladas:
\begin{align}
\begin{cases} \partial_{t}r_{1}(t) = \omega r_{2}(t), \\ \partial_{t}r_{2}(t) = -\omega r_{1}(t).  \end{cases}
\end{align}
Do qual obtemos o sistema de eqs. diferenciais lineares de segunda ordem desacopladas:
\begin{align}
\begin{cases} \partial_{tt}r_{1}(t) = \omega \partial_{t}r_{2}(t) = -\omega^{2}r_{1}(t), \\ \partial_{tt}r_{2}(t) = -\omega \partial_{t}r_{1}(t) = -\omega^{2}r_{2}(t), \end{cases}
\end{align}
cuja solução pode ser escrita como:
\begin{align}
\begin{cases} r_{1}(t) = r_{1}\cos\omega t + r_{2}\sin\omega t, \\ r_{2}(t) = -r_{1}\sin\omega t + r_{2}\cos\omega t,\\ r_{3}(t)=r_{3}. \end{cases}
\end{align}
Matriz densidade evoluída:
\begin{align}
\rho_{t} & = \begin{bmatrix} 1+r_{3}(t) & r_{1}(t)-ir_{2}(t) \\ r_{1}(t)+ir_{2}(t) & 1+r_{3}(t) \end{bmatrix} \\
         & = \vdots \\
         & = \begin{bmatrix} 1+r_{3} & (r_{1}-ir_{2})e^{i\omega t} \\ (r_{1}+ir_{2})e^{-i\omega t} & 1+r_{3} \end{bmatrix}.
\end{align}

__Exercício:__ Considere o mesmo sistema do último exemplo, mas obtenha o estado evoluído usando $\rho_{t}=U_{t}\rho U_{t}^{\dagger}$ com $U_{t}=e^{-iHt/\hbar}$.

## Entropia
Existe uma diferença fundamental entre a incerteza clássica associada ao operador densidade e a incerteza quântica associada com superposição. Esses dois tipos de incerteza podem coexistir. A seguir definiremos funções que podem ser utilizadas para quantificar essas incertezas distintivamente.

### Entropia linear (EL)
A _pureza_ de um operador densidade é definida com
\begin{equation}
P(\rho) = \mathrm{Tr}(\rho^{2}).
\end{equation}
A EL é definida como:
\begin{equation}
S_{L}(\rho) = 1-P(\rho).
\end{equation}
__Exercício:__ Verifique as igualdades em: $S_{L}(\mathbb{I}/d) = (d-1)/d \ge S_{L}(\rho) \ge 0 = S_{L}(|\psi\rangle)$.

### Entropia de von Neumann (EN)
\begin{equation}
S_{vN}(\rho) = -\mathrm{Tr}(\rho\log_{2}\rho) = H(\{r_{j}\}),
\end{equation}
onde $H(\{r_{j}\})=-\sum_{j}r_{j}\log_{2}r_{j}$ é a _entropia de Shannon_ e $r_{j}$ são os autovalores de $\rho$. <br>
_OBS:_ Usamos $\log_{2}$ para medir entropia em bits. <br>
_OBS:_ Em termodinâmica usa-se $\ln=\log_{e}$ no lugar de $\log_{2}$ e a expressão da entropia é multiplicada por $k_{B}$, a constante de Boltzmann. <br>
_OBS:_ Ao calcular $S_{vN}$ usamos $\lim_{x\rightarrow 0} x\log x = 0$ para fazer $0\log 0 = 0$. <br>
__Exercício:__ Verifique as igualdades em: $S_{vN}(\mathbb{I}/d) = \log_{2}d \ge S_{vN}(\rho) \ge 0 = S_{vN}(|\psi\rangle)$.

### Entropia de Wigner-Yanase (EWY)
Enquanto as EL e EN quantificam incerteza clássica, a EWY quantifica incerteza quântica em relação a um observável $O$ para um sistema no estado $\rho$:
\begin{align}
S_{wy}(O|\rho) & = -\frac{1}{2}\mathrm{Tr}\left([\sqrt{\rho},O]^{2}\right) = -\frac{1}{2}\mathrm{Tr}\left((\sqrt{\rho}O-O\sqrt{\rho})^{2}\right) = -\frac{1}{2}\mathrm{Tr}\left( (\sqrt{\rho}O-O\sqrt{\rho})(\sqrt{\rho}O-O\sqrt{\rho}) \right) \\
               & = -\frac{1}{2}\mathrm{Tr}\left( \sqrt{\rho}O\sqrt{\rho}O -O\sqrt{\rho}\sqrt{\rho}O - \sqrt{\rho}OO\sqrt{\rho} +O\sqrt{\rho}O\sqrt{\rho} \right) \\
               & = -\frac{1}{2}\left( \mathrm{Tr}(\sqrt{\rho}O\sqrt{\rho}O) - \mathrm{Tr}(O\rho O) - \mathrm{Tr}(\sqrt{\rho}O^{2}\sqrt{\rho}) + \mathrm{Tr}(O\sqrt{\rho}O\sqrt{\rho}) \right) \\
               & = \mathrm{Tr}(\rho O^{2}) - \mathrm{Tr}(\sqrt{\rho}O\sqrt{\rho}O) = \langle O^{2}\rangle_{\rho} - \mathrm{Tr}\left((O\sqrt{\rho})^{2}\right).
\end{align}
_OBS:_ Note que se $\rho$ e $O$ comutam, a EWY é nula, mesmo que nossa incerteza clássica sobre o estado do sistema seja positiva. <br>
__Exercício:__ Verifique que para estados puros $S_{wy}(O|\rho)=\mathrm{Var}(O)_{\psi}$.

### Exemplo
_OBS:_ O comprimento do vetor de Bloch não muda por transformações unitárias. <br>
Para verificar essa informação, considera $\rho'=U\rho U^{\dagger}$. Para sistemas de 2 níveis
\begin{align}
P(\rho) & = \mathrm{Tr} \frac{1}{2}\left(\sigma_{0} + \sum_{j=1}^{3}r_{j}\sigma_{j}\right)\frac{1}{2}\left(\sigma_{0} + \sum_{k=1}^{3}r_{k}\sigma_{k}\right) = \frac{1}{4}\mathrm{Tr}\left(\sigma_{0} + 2\sum_{j=1}^{3}r_{j}\sigma_{j} + \sum_{j,k=1}^{3}r_{j}r_{k}\sigma_{j}\sigma_{k} \right) \\
        & = \frac{1}{4}\left(2 + 0 + \sum_{j,k=1}^{3}r_{j}r_{k}\mathrm{Tr}(\sigma_{j}\sigma_{k}) \right) = \frac{1}{4}\left(2 + \sum_{j,k=1}^{3}r_{j}r_{k}2\delta_{jk} \right) \\
        & = \frac{1}{2}\left(1 + r^{2} \right).
\end{align}
Pela invariância da função traço sob transformações unitárias (mudança de base),
\begin{align}
P(\rho') & = \frac{1}{2}\left(1 + r'^{2} \right) = \mathrm{Tr}(U\rho U^{\dagger}U\rho U^{\dagger}) = \mathrm{Tr}(U\rho\rho U^{\dagger}) \\
         & = \mathrm{Tr}(\rho^{2}).
\end{align}
Portanto $r'=r$. Ou seja, tranformações unitárias somente giram o vetor de Bloch em $\mathbb{R}^{3}$.

Vamos calcular aquelas três entropias para o estado geral de um sistema de 2 níveis mostrado acima. Da expressão para a pureza vem que
\begin{equation}
S_{L}(\rho) = \frac{1-r^{2}}{2}.
\end{equation}
Já, dos autovalores de $\rho$, a EN fica
\begin{align}
S_{vN}(\rho) & = -\frac{1+r}{2}\log_{2}\frac{1+r}{2} - \frac{1-r}{2}\log_{2}\frac{1-r}{2} \\
             & = \vdots \\
             & = 1 - 2^{-1}[(1+r)\log_{2}(1+r) + (1-r)\log_{2}(1-r)].
\end{align}


Para obter $S_{wy}$ consideraremos como observável $\sigma_{z}$. Nesse caso, como $\sigma_{z}^{2}=\sigma_{0}$, fazendo $\rho=r_{+}|r_{+}\rangle\langle r_{+}|+r_{-}|r_{-}\rangle\langle r_{-}|$ com $r_{\pm}=(1\pm r)/2$ e $|r_{+}\rangle = \cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle$ e $|r_{-}\rangle = \sin(\theta/2)|z_{+}\rangle - e^{i\phi}\cos(\theta/2)|z_{-}\rangle$. Assim, usando $\sigma_{z}|r_{+}\rangle = \cos(\theta/2)|z_{+}\rangle - e^{i\phi}\sin(\theta/2)|z_{-}\rangle$,  $\sigma_{z}|r_{-}\rangle=\sin(\theta/2)|z_{+}\rangle + e^{i\phi}\cos(\theta/2)|z_{-}\rangle$ e $\mathrm{Tr}(A|\psi\rangle\langle\psi|)=\langle\psi|A|\psi\rangle$ obteremos
\begin{align}
S_{wy}(\sigma_{z}|\rho) & = \mathrm{Tr}(\rho) - \mathrm{Tr}\left((\sigma_{z}\sqrt{\rho})^{2}\right) = 1 - \mathrm{Tr}\left((\sigma_{z}(\sqrt{r_{+}}|r_{+}\rangle\langle r_{+}|+\sqrt{r_{-}}|r_{-}\rangle\langle r_{-}|))^{2}\right) \\
                        & = 1 - \mathrm{Tr}(\sqrt{r_{+}}\sigma_{z}|r_{+}\rangle\langle r_{+}| + \sqrt{r_{-}}\sigma_{z}|r_{-}\rangle\langle r_{-}|)^{2} \\
                        & = 1 - \mathrm{Tr}\left(r_{+}\sigma_{z}|r_{+}\rangle\langle r_{+}|\sigma_{z}|r_{+}\rangle\langle r_{+}| + \sqrt{r_{+}r_{-}}\sigma_{z}|r_{+}\rangle\langle r_{+}|\sigma_{z}|r_{-}\rangle\langle r_{-}| + \sqrt{r_{+}r_{-}}\sigma_{z}|r_{-}\rangle\langle r_{-}|\sigma_{z}|r_{+}\rangle\langle r_{+}| + r_{-}\sigma_{z}|r_{-}\rangle\langle r_{-}|\sigma_{z}|r_{-}\rangle\langle r_{-}| \right) \\
                        & = 1 - r_{+}\langle r_{+}|\sigma_{z}|r_{+}\rangle^{2} - 2\sqrt{r_{+}r_{-}}|\langle r_{-}|\sigma_{z}|r_{+}\rangle|^{2} - r_{-}\langle r_{-}|\sigma_{z}|r_{-}\rangle^{2} \\
                        & = 1 - r_{+}(\cos^{2}(\theta/2)-\sin^{2}(\theta/2))^{2} - 2\sqrt{r_{+}r_{-}}|2\sin(\theta/2)\cos(\theta/2)|^{2} - r_{-}(\sin^{2}(\theta/2)-\cos^{2}(\theta/2))^{2} \\
                        & = 1 - 2^{-1}(1+r)\cos^{2}\theta - \sqrt{1-r^{2}}\sin^{2}\theta - 2^{-1}(1-r)\cos^{2}\theta  = 1 - \cos^{2}\theta - \sqrt{1-r^{2}}\sin^{2}\theta \\
                        & = (1-\sqrt{1-r^{2}})\sin^{2}\theta.
\end{align}

```python
def entropies(n):
    th = n*np.pi/8
    N = 100
    dr = (1.01-0)/N
    rv = np.zeros(N)
    P = np.zeros(N)
    Sl = np.zeros(N)
    Sn = np.zeros(N)
    Swy = np.zeros(N)
    #th = entropies(th)
    r = -dr
    for j in range(0,N):
        r += dr
        rv[j] = r
        P[j] = (1+r**2)/2
        Sl[j] = 1-P[j]
        Sn[j] = 1-(1/2)*((1+r)*log(1+r,2) + (1-r)*log(1-r,2))
        Swy[j] = (1-sqrt(1-r**2))*(sin(th))**2
    plt.xlabel(r'$r$')
    plt.plot(rv, P, label=r'$P$')
    plt.plot(rv, Sl, label=r'$S_{L}$')
    plt.plot(rv, Sn, label=r'$S_{vN}$')
    plt.plot(rv, Swy, label=r'$S_{WY}(\theta=\pi/4)$')
    axes = plt.gca()
    axes.set_xlim([0,1])
    axes.set_ylim([-0.1,1.1])
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.show()
```

```python
interactive(entropies, n=(0,8))
```

## Estado térmico de Gibbs
Emaranhamento do sistema com seu ambiente, que possui temperatura $T$ e é "muito grande", fará com que o estado de equilíbrio do primeiro tenha a seguinte forma [arXiv:quant-ph/0511225]:
\begin{equation}
\rho_{th} = \frac{e^{-\beta H}}{Z} = \sum_{j} \frac{e^{-\beta E_{j}}}{Z}|E_{j}\rangle\langle E_{j}| = \sum_{j} Pr(E_{j})|E_{j}\rangle\langle E_{j}|,
\end{equation}
em que $\beta = 1/k_{B}T$, $H=\sum_{j}E_{j}|E_{j}\rangle\langle E_{j}|$ é o Hamiltoniano do sistema e a normalização (para que $\mathrm{Tr}(\rho_{th})=1$) é a função de partição:
\begin{equation}
Z = \mathrm{Tr}\left(e^{-\beta H}\right) = \sum_{j}e^{-\beta E_{j}}.
\end{equation}

Nesse caso,
\begin{align}
S_{th}(\rho_{th}) & = k_{B}\ln 2 S_{vN}(\rho_{th}) = - k_{B}\mathrm{Tr}(\rho_{th}\ln\rho_{th}) = -k_{B}H\left(\left\{\frac{e^{-\beta E_{j}}}{Z}\right\}\right) = - k_{B}\sum_{j}\frac{e^{-\beta E_{j}}}{Z}\ln\frac{e^{-\beta E_{j}}}{Z} \\
& = - k_{B}\sum_{j}\frac{e^{-\beta E_{j}}}{Z}(\ln e^{-\beta E_{j}} - \ln Z) = - k_{B}\left(-\beta\sum_{j}\frac{e^{-\beta E_{j}}}{Z} E_{j}-\ln Z\sum_{j}\frac{e^{-\beta E_{j}}}{Z}\right) \\
&= k_{B}\frac{1}{k_{B}T} \langle H\rangle + k_{B}\ln Z \\
\end{align}
ou
\begin{equation}
-k_{B}T\ln Z =: F = \langle H\rangle - TS_{th}.
\end{equation}

### Exemplo
Considere um spin 1/2 com Hamiltoniano $H=-(\hbar\omega/2)\sigma_{z} = E_{+}|z_{+}\rangle\langle z_{+}|+E_{-}|z_{-}\rangle\langle z_{-}|$. Primeiro calculamos a função de partição
\begin{align}
Z & = \mathrm{Tr}(e^{-\beta H}) = \sum_{j}e^{-\beta E_{j}} = e^{-\beta E_{+}} + e^{-\beta E_{-}} = e^{-\beta (-\hbar\omega/2)} + e^{-\beta(\hbar\omega/2)} \\
  & = 2\cosh\frac{\beta\hbar\omega}{2}.
\end{align}
Energia média
\begin{align}
\langle H\rangle & = E_{+}Pr(E_{+}) + E_{-}Pr(E_{-}) = \frac{-\hbar\omega}{2}\frac{e^{\beta\hbar\omega/2}}{Z} + \frac{\hbar\omega}{2}\frac{e^{-\beta\hbar\omega/2}}{Z} = \cdots \\
                 & = -\frac{\hbar\omega}{2}\tanh\frac{\beta\hbar\omega}{2}.
\end{align}
Entropia
\begin{align}
\frac{S_{th}}{k_{B}} & = \frac{\langle H\rangle}{T} + k_{B}\ln Z = \frac{-\frac{\hbar\omega}{2}\tanh\frac{\beta\hbar\omega}{2}}{k_{B}T} + \ln\left(2\cosh\frac{\beta\hbar\omega}{2}\right) \\
                     & = -\frac{\beta\hbar\omega}{2}\tanh\frac{\beta\hbar\omega}{2} + \ln\left(2\cosh\frac{\beta\hbar\omega}{2}\right).
\end{align}

```python
def thermal(ymax):
    N = 150
    db = (6-0)/N
    bv = np.zeros(N)
    Pp = np.zeros(N)
    Z = np.zeros(N)
    Em = np.zeros(N)
    Sth = np.zeros(N)
    bhw = -db
    for j in range(0,N):
        bhw += db
        bv[j] = bhw
        Z[j] = 2*cosh(bhw/2)
        Pp[j] = exp(bhw/2)/Z[j]
        Em[j] = -(1/2)*tanh(bhw*1/2) # A energia está em unidades de \hbar\omega
        Sth[j] = (-(bhw/2)*tanh(bhw/2) + ln(2*cosh(bhw/2)))/ln(2)
    plt.xlabel(r'$\beta\hbar\omega = \hbar\omega/k_{B}T$')
    plt.plot(bv, Z, label=r'$Z$')
    plt.plot(bv, Pp, label=r'$Pr(E_{+})$')
    plt.plot(bv, Em, label=r'$\langle H\rangle$')
    plt.plot(bv, Sth, label=r'$S_{th}/(k_{B}\ln2)$')
    axes = plt.gca()
    axes.set_xlim([0,6])
    axes.set_ylim([-1,ymax])
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.show()
```

```python
interactive(thermal, ymax = (3,10))
```
