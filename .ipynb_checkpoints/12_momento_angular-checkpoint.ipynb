# Momento angular orbital
Obtemos o momento angular quântico quantizando a expressão clássica:
\begin{align}
& \vec{L} = \vec{X}\mathrm{x}\vec{P} \\
& = \sum_{j=1}^{3}\hat{e}_{j}(\vec{X}\mathrm{x}\vec{P})_{j} \\
& = \sum_{j=1}^{3}\hat{e}_{j}\sum_{k,l=1}^{3}\varepsilon_{jkl}X_{k}P_{l} \\
& = \sum_{j=1}^{3}\hat{e}_{j}L_{j}, \\
& \therefore L_{j}=\sum_{k,l=1}^{3}\varepsilon_{jkl}X_{k}P_{l}.
\end{align}

## Relação de comutação para as componentes do momento angular (orbital)
Usando novamente $[A,BC]=B[A,C]+[A,B]C$ e $[AB,C] = A[B,C]+[A,C]B$, teremos
\begin{align}
[L_{j},L_{k}] & = \left[\sum_{l,m=1}^{3}\varepsilon_{jlm}X_{l}P_{m} , \sum_{p,q=1}^{3}\varepsilon_{kpq}X_{p}P_{q} \right] \\
& = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left[X_{l}P_{m} , X_{p}P_{q} \right] \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(X_{l}[P_{m} , X_{p}P_{q}] + [X_{l} , X_{p}P_{q}]P_{m} \right) \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(X_{l}X_{p}[P_{m},P_{q}] + X_{l}[P_{m},X_{p}]P_{q} + X_{p}[X_{l},P_{q}]P_{m} + [X_{l} , X_{p}]P_{q}P_{m} \right) \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(\mathbb{0} + X_{l}(-i\hbar\mathbb{1}\delta_{mp})P_{q} + X_{p}(i\hbar\mathbb{1}\delta_{lq})P_{m} + \mathbb{0} \right) \\
              & = -i\hbar\sum_{l,m,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kmq}X_{l}P_{q} + i\hbar\sum_{l,m,p=1}^{3}\varepsilon_{jlm}\varepsilon_{kpl}X_{p}P_{m} \\
\end{align}

Podemos verificar a seguinte identidade:
\begin{equation}
\sum_{j=1}^{3}\varepsilon_{jkl}\varepsilon_{jmn} = \delta_{km}\delta_{ln} - \delta_{kn}\delta_{lm}.
\end{equation}

```python
ss = 0 # verificação da última igualdade
for k in range(1,4):
    for l in range(1,4):
        for m in range(1,4):
            for n in range(1,4):
                s = 0
                for j in range(1,4):
                    s += LeviCivita(j,k,l)*LeviCivita(j,m,n)
                ss += abs(s-(KroneckerDelta(k,m)*KroneckerDelta(l,n)-KroneckerDelta(k,n)*KroneckerDelta(l,m)))
                #print(ss)
```

Obtemos assim
\begin{align}
[L_{j},L_{k}] & = i\hbar\sum_{l,q=1}^{3}\sum_{m=1}^{3}\varepsilon_{mjl}\varepsilon_{mkq}X_{l}P_{q} - i\hbar\sum_{m,p=1}^{3}\sum_{l=1}^{3}\varepsilon_{ljm}\varepsilon_{lkp}X_{p}P_{m} \\
              & = i\hbar\sum_{l,q=1}^{3}(\delta_{jk}\delta_{lq}-\delta_{jq}\delta_{lk})X_{l}P_{q} - i\hbar\sum_{m,p=1}^{3}(\delta_{jk}\delta_{mp}-\delta_{jp}\delta_{mk})X_{p}P_{m}.
\end{align}
Como para $j=k$ temos $[L_{j},L_{k}]=\mathbb{O}$, vamos considerar a relação acima com $j\ne k\Rightarrow \delta_{jk}=0$. Assim
\begin{align}
[L_{j},L_{k}] & = -i\hbar\sum_{l,q=1}^{3}\delta_{jq}\delta_{lk}X_{l}P_{q} + i\hbar\sum_{m,p=1}^{3}\delta_{jp}\delta_{mk}X_{p}P_{m} \\
              & = i\hbar(-X_{k}P_{j} + X_{j}P_{k}).
\end{align}

Substituindo explicitamente os índices, vem que
\begin{align}
[L_{1},L_{2}] & = i\hbar(X_{1}P_{2}-X_{2}P_{1}) \\
& = i\hbar\sum_{j,k}\varepsilon_{3jk}X_{j}P_{k} \\ 
& = i\hbar L_{3} \\
& =  i\hbar\varepsilon_{123}L_{3}.
\end{align}
Analogamente, verificar-se que 
\begin{align}
[L_{2},L_{3}] & = i\hbar(X_{2}P_{3} - X_{3}P_{2}) \\
& = i\hbar(\epsilon_{123}X_{2}P_{3} + \epsilon_{132}X_{3}P_{2}) \\
& = i\hbar L_{1} \\
& = i\hbar\varepsilon_{231}L_{1}, \\
[L_{1},L_{3}] & = i\hbar(X_{1}P_{3} - X_{3}P_{1}) \\
& = i\hbar(-\epsilon_{213}X_{1}P_{3} - \epsilon_{231}X_{3}P_{1}) \\
& = -i\hbar L_{2} \\
& = i\hbar\varepsilon_{132}L_{2}.
\end{align}
Com isso, a importante relação de comutação para as componentes do momento angular pode ser escrita em uma forma geral:
\begin{align}
[L_{j},L_{k}] & = i\hbar\varepsilon_{jkl}L_{l} \\
& = i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l}.
\end{align}
Essa relação de comutação defina a chamada __álgebra do momento angular__.

## Autovalores do momento angular
Aqui, consideremos qualquer operador $\vec{L}$ cujas componentes satisfazem a relação de comutação $[L_{j},L_{k}] = i\hbar\varepsilon_{jkl}L_{l}$. O método de diagonalização que usaremos aqui é semelhante àquele utilizado na solução do oscilador harmônico. Primeiramente verifiquemos que o quadrado do momento angular comuta com todas as suas componentes:
\begin{align}
[L^{2},L_{k}] & = \left[\sum_{j=1}^{3}L_{j}^{2},L_{k}\right] \\
& = \sum_{j=1}^{3}[L_{j}L_{j},L_{k}] \\
& = \sum_{j=1}^{3}(L_{j}[L_{j},L_{k}] + [L_{j},L_{k}]L_{j}) \\
& = \sum_{j=1}^{3}\left(L_{j}i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l} + i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l}L_{j}\right) \\
& = i\hbar\sum_{j,l=1}^{3}\varepsilon_{jkl}(L_{j}L_{l} + L_{l}L_{j}) \\
& = i\hbar\sum_{j,l=1}^{3}\varepsilon_{jkl}\{L_{j},L_{l}\}.
\end{align}
Considerando as componentes explicitamente, temos por exemplo que
\begin{align}
[L^{2},L_{1}] & = i\hbar(\varepsilon_{213}\{L_{2},L_{3}\} + \varepsilon_{312}\{L_{3},L_{2}\}) \\
& = i\hbar\{L_{2},L_{3}\}(\varepsilon_{213} + \varepsilon_{312}) \\ 
& = i\hbar\{L_{2},L_{3}\}(\varepsilon_{213} - \varepsilon_{213}) \\
& = 0,
\end{align}
onde usamos 
$$\{L_{j},L_{k}\}=\{L_{k},L_{j}\}.$$
Analogamente, pode-se verificar também que 
$$[L^{2},L_{2}]=[L^{2},L_{3}]=0.$$
Com isso vemos que
\begin{equation}
[L^{2},L_{k}] = 0  \text{ para } k=1,2,3.
\end{equation}

Seguindo, consideremos que $L^{2}$ compartilha a base de autovetores $|l,m\rangle$ com $L_{3}$ e
\begin{cases}
L^{2}|l,m\rangle = \lambda|l,m\rangle \\
L_{3}|l,m\rangle = \hbar m|l,m\rangle.
\end{cases}
Vale observar que aqui escolhemos a direção $z$ com a __direção de quantização__, mas qualquer outra escolha levaria aos mesmos resultados observacionais que esta.
Vamos verificar que $\lambda = \hbar^{2}l(l+1)$ com $l=n/2$ para $n\in\mathbb{N}$ e $m=-l,-l+1,\cdots,-1,0,1\cdots,l-1,l$. Para isso precisaremos dos __operadores escada__:
\begin{equation}
L_{\pm} := L_{1} \pm iL_{2}.
\end{equation}
Como
\begin{equation}
[L^{2},L_{\pm}] = [L^{2},L_{1}] \pm i[L^{2},L_{2}] = 0, 
\end{equation}
vemos que os vetores $L_{\pm}|l,m\rangle$ também são autovetores de $L^{2}$ correspondentes aos mesmos autovalores $\lambda$ que $|l,m\rangle$, i.e.,
\begin{equation}
L^{2}(L_{\pm}|l,m\rangle) = \lambda(L_{\pm}|l,m\rangle).
\end{equation}
Por outro lado, os operadores escada não comutam com as componentes do momento angular. Por exemplo
\begin{align}
[L_{3},L_{\pm}] & = [L_{3},L_{1}] \pm i[L_{3},L_{2}] \\
& = i\hbar\varepsilon_{312}L_{2} \pm i i\hbar\varepsilon_{321}L_{1} \\
&= i\hbar L_{2} \pm \hbar L_{1} \\
&= \pm\hbar(L_{1} \pm iL_{2}) \\
& = \pm\hbar L_{\pm}.
\end{align}
Por outro lado, vemos que
\begin{align}
L_{3}(L_{\pm}|l,m\rangle) & = (L_{3}L_{\pm})|l,m\rangle \\
& = (\pm\hbar L_{\pm} + L_{\pm}L_{3})|l,m\rangle \\
&= \pm\hbar L_{\pm}|l,m\rangle + L_{\pm}L_{3}|l,m\rangle \\
& = \pm\hbar L_{\pm}|l,m\rangle + L_{\pm}\hbar m|l,m\rangle \\
& = \hbar( m \pm 1)(L_{\pm}|l,m\rangle).
\end{align}

Ou seja, os vetores $L_{\pm}|l,m\rangle$ também são autovetores de $L_{3}$, mas com autovalores aumentados ou diminuídos de $\hbar$ em relação àquele associado com $|l,m\rangle$. Agora, como 
$$L^{2} = L_{1}^{2}+L_{2}^{2}+L_{3}^{2},$$
não poderemos aumentar nem diminuir indefinidamente os autovalores de $L_{3}$ através da aplicação dos operadores escada. Por isso, para os valores definidos por $m:=\pm l$ a ação dos operadores escada nos retorna o vetor nulo:
\begin{equation}
L_{\pm}|l,m =\pm l\rangle := |\oslash\rangle.
\end{equation}

Note que como $m_{\min}=-l$ e $m_{\max}=l$ e $m$ aumenta ou diminiu de 1 em 1, para um dado $l$, os possíveis valores de $m$ são:
\begin{equation}
m=-l,-l+1,\cdots,l-1,l.
\end{equation}

Continuando, vamos escrever $\lambda$ em termos de $l$. Para isso consideremos
\begin{align}
L_{\pm}L_{\mp} & = (L_{1} \pm iL_{2})(L_{1} \mp iL_{2}) \\
& = L_{1}^{2} \mp iL_{1}L_{2} \pm iL_{2}L_{1} + L_{2}^{2}  \\ 
& = L_{1}^{2} + L_{2}^{2} + L_{3}^{2} - L_{3}^{2} \mp i[L_{1},L_{2}]  \\
& = L^{2} - L_{3}^{2} \mp ii\hbar\varepsilon_{123}L_{3} \\
& = L^{2} - L_{3}^{2} \pm\hbar L_{3}.
\end{align}
Com isso, podemos escrever
\begin{align}
L^{2}|l,m=l\rangle & = (L_{-}L_{+} + L_{3}^{2} + \hbar L_{3})|l,l\rangle  \\ 
&= L_{-}L_{+}|l,l\rangle + L_{3}^{2}|l,l\rangle + \hbar L_{3}|l,l\rangle  \\
& = |\oslash\rangle + \hbar^{2}l^{2}|l,l\rangle + \hbar\hbar l|l,l\rangle \\
                 & = \hbar^{2}l(l+1)|l,l\rangle  \\
                 & = \lambda|l,l\rangle.
\end{align}
Analogamente, pode-se verificar que $L^2|l,-l\rangle=\hbar^2 l(l+1)|l,-l\rangle$.

### Domínio de $l$
Consideremos que os autovalores de $L_{3}$ são, por definição, $\hbar m$ e que vimos que se começamos no estado $|l,m=-l\rangle$ e aplicamos $L_{+}$ um número $n$ de vezes, aumentaremos o valor de $m$ de uma unidade em cada aplicação de $L_{+}$, até obtermos o estado $|l,m=l\rangle$, i.e.,
\begin{equation}
|l,m=l\rangle := L_{+}^{n}|l,m=-l\rangle \text{, com } n\in\mathbb{N}.
\end{equation}
Como $n\in\mathbb{N}$ teremos que 
\begin{align}
& l=-l+n\ \therefore\ 2l=n \ \therefore\ l = \frac{n}{2} \\
& \therefore l= 0, \frac{1}{2}, 1, \frac{3}{2}, 2, \frac{5}{2},\cdots.
\end{align}

### Ação dos operadores escada na base $|l,m\rangle$
Como os operadores escada aumentam ou diminuem o número quântico $m$ de uma unidade, teremos e.g.
\begin{equation}
L_{+}|l,m\rangle \propto |l,m+1\rangle =: c|l,m+1\rangle,
\end{equation}
onde introduzimos a constate $c$ para transformar a proporcionalidade em igualdade. Olhemos agora para o quadrado da norma desse vetor
\begin{align}
\langle l,m|L_{+}^{\dagger}L_{+}|l,m\rangle & = \langle l,m|L_{-}L_{+}|l,m\rangle \\
& = \langle l,m|(L^{2} - L_{3}^{2} -\hbar L_{3})|l,m\rangle \\
& = \langle l,m|\left(\hbar^{2}l(l+1)-\hbar^{2}m^{2}-\hbar^{2}m\right)|l,m\rangle \\
&= \left(\hbar^{2}l(l+1)-\hbar^{2}m^{2}-\hbar^{2}m\right)\langle l,m|l,m\rangle \\
& = \hbar^{2}[l(l+1) - m(m+1)]  \\
& = \langle l,m+1|c^{*}c|l,m+1\rangle \\
& = |c|^{2}.
\end{align}
Usando $c\in\mathbb{R}$ obtemos
\begin{equation}
L_{+}|l,m\rangle = \hbar\sqrt{l(l+1)-m(m+1)}|l,m+1\rangle.
\end{equation}
Pode-se mostrar também que
\begin{equation}
L_{-}|l,m\rangle = \hbar\sqrt{l(l+1)-m(m-1)}|l,m-1\rangle.
\end{equation}

# Soma de momentos angulares
Na Mecânica Quântica de Schrödinger, o momento angular intrínseco, o spin, é descrito por operadores que satisfazem a álgebra do momento angular. Na sequência, consideraremos o caso para o qual, para descrevermos o sistema, precisamos somar 2 ou mais momentos angulares. Pode ser o momento angular orbital e o spin de um elétron, pode ser os spins de dois elétrons, ou qualquer outra combinação de 2 momentos angulares.

Consideremos os momentos angulares $\vec{J}_{s}$ com $s=a,b$, ou seja, consideremos dois vetores cujas componentes satisfazem a álgebra do momento angular:
\begin{equation}
[J_{j}^{s},J_{k}^{s}]=i\hbar\varepsilon_{jkl}J_{l}^{s} \text{ para } j,k=1,2,3.
\end{equation}
Para esses dois graus de liberdade, a soma desses momentos angulares é representada por um operador Hermitiano que atua no espaço de Hilbert composto $\mathcal{H}_{a}\otimes\mathcal{H}_{b}$ e que é escrito como segue:
\begin{align}
\vec{J} & := \vec{J}_{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes\vec{J}_{b} \\
& = \sum_{i=1}^{3}\hat{e}_{i}\left(J_{i}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{i}^{b}\right) \\
& = \sum_{i=1}^{3}\hat{e}_{i}J_{j}.
\end{align}

Para os conjuntos de observáveis
\begin{equation}
\{J_{a}^{2},J_{3}^{a},J_{b}^{2},J_{3}^{b}\} \text{ e } \{J_{a}^{2},J_{b}^{2},J^{2},J_{3}\},
\end{equation}
verificaremos na sequência que os observáveis dentro de cada um dos conjuntos formam um conjunto de observáveis mutuamente compatíveis, lembrando que observáveis mutuamente compatíveis comutam e portanto compartilham a mesma base de autovetores. Bom, da álgebra de momento angular já temos que 
$$[J_{s}^{2},J_{3}^{s}]=0.$$
Para os cálculos que seguem, primeiro usamos 
$$(A\otimes B)(C\otimes D) = AC\otimes BD,$$
que vale para quaisquer operadores $A,C:\mathcal{H}_{a}\rightarrow\mathcal{H}_{a}$ e $B,D:\mathcal{H}_{b}\rightarrow\mathcal{H}_{b}$, para ver que 
\begin{align}
[A\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes B] & = (A\otimes\mathbb{I}_{b})(\mathbb{I}_{a}\otimes B) - (\mathbb{I}_{a}\otimes B)(A\otimes\mathbb{I}_{b}) \\
&= (A\mathbb{I}_{a}\otimes\mathbb{I}_{b}B) - (\mathbb{I}_{a}A\otimes B\mathbb{I}_{b}) \\
& = A\otimes B - A\otimes B \\
& = 0.
\end{align}
Esse resultado pode ser aplicado diretamente a
\begin{align}
& [J_{a}^{2},J_{b}^{2}] = [J_{a}^{2}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{b}^{2}] = 0, \\
& [J_{a}^{2},J_{3}^{b}] = [J_{a}^{2}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{3}^{b}] = 0, \\
& [J_{b}^{2},J_{3}^{a}] = [\mathbb{I}_{a}\otimes J_{b}^{2},J_{3}^{a}\otimes\mathbb{I}_{b}] = 0, \\
& [J_{3}^{a},J_{3}^{b}] = [J_{3}^{a}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{3}^{b}] = 0.
\end{align}
Verificamos assim que todos os operadores do conjunto $\{J_{a}^{2},J_{3}^{a},J_{b}^{2},J_{3}^{b}\}$ comutam.

Agora vamos para o segundo conjunto de operadores. Aqui começamos verificando que $\vec{J}$ é de fato um momento angular, ou seja, verifiquemos que suas componentes satisfazem a álgebra do momento angular:
\begin{align}
[J_{j},J_{k}] & = [J_{j}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{j}^{b}, J_{k}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{k}^{b}] \\
              & = [J_{j}^{a}\otimes\mathbb{I}_{b},J_{k}^{a}\otimes\mathbb{I}_{b}] + [J_{j}^{a}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{k}^{b}] + [\mathbb{I}_{a}\otimes J_{j}^{b},J_{k}^{a}\otimes\mathbb{I}_{b}] + [\mathbb{I}_{a}\otimes J_{j}^{b},\mathbb{I}_{a}\otimes J_{k}^{b}] \\
              & = [J_{j}^{a},J_{k}^{a}]\otimes\mathbb{I}_{b} + 0 + 0 + \mathbb{I}_{a}\otimes[J_{j}^{b},J_{k}^{b}] \\
              & = i\hbar\varepsilon_{jkl}J_{l}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes i\hbar\varepsilon_{jkl}J_{l}^{b} \\
              & = i\hbar\varepsilon_{jkl}\left(J_{l}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{l}^{b}\right)  \\
              & = i\hbar\varepsilon_{jkl}J_{l}.
\end{align}
Acima usamos que
$$[A\otimes\mathbb{I}_B,A'\otimes\mathbb{I}_B] = [A,A']\otimes\mathbb{I}_B$$
e
$$[\mathbb{I}_A\otimes B,\mathbb{I}_A\otimes B'] = \mathbb{I}_A\otimes[B,B'].$$
Assim, pela mesma demonstração feita no caso do momento angular orbital, temos aqui que
\begin{equation}
[J^{2},J_{j}] = 0 \text{ para }  j=1,2,3.
\end{equation}
 Desse resultado segue que $[J^{2},J_{3}]=0$. Pode-se verificar que
\begin{align}
J^{2} & = \vec{J}\cdot\vec{J} \\
& =  J_{a}^{2}\otimes\mathbb{I}_{b} +  \mathbb{I}_{a}\otimes J_{b}^{2} + 2\sum_{j=1}^{3}J_{j}^{a}\otimes J_{j}^{b}.
\end{align}

Usando
\begin{equation}
[A\otimes B,A'\otimes\mathbb{I}] = [A,A']\otimes B
\end{equation}
vemos que
\begin{align}
[J^{2},J_{a}^{2}] & = [J^{2},J_{a}^{2}\otimes\mathbb{I}_{b}] \\
                  & = [J_{a}^{2}\otimes\mathbb{I}_{b},J_{a}^{2}\otimes\mathbb{I}_{b}] + [\mathbb{I}_{a}\otimes J_{b}^{2},J_{a}^{2}\otimes\mathbb{I}_{b}] + [2\sum_{j=1}^{3}J_{j}^{a}\otimes J_{j}^{b},J_{a}^{2}\otimes\mathbb{I}_{b}] \\
                  & = 0 + 0 + 2\sum_{j=1}^{3}[J_{j}^{a},J_{a}^{2}]\otimes J_{j}^{b} = 0.
\end{align}
Do mesmo modo, pode-se verificar que
\begin{equation}
[J^{2},J_{b}^{2}] = 0.
\end{equation}
Nos falta ainda olhar para
\begin{align}
[J_{a}^{2},J_{3}] & = [J_{a}^{2}\otimes\mathbb{I}_{b},J_{3}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{3}^{b}] \\ 
& = [J_{a}^{2}\otimes\mathbb{I}_{b},J_{3}^{a}\otimes\mathbb{I}_{b}] + [J_{a}^{2}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{3}^{b}] \\
& = [J_{a}^{2},J_{3}^{a}]\otimes\mathbb{I}_{b} + 0 \\
                  & = 0.
\end{align}
Analogamente, pode-se verificar que
\begin{equation}
[J_{b}^{2},J_{3}] = 0.
\end{equation}

Com isso, verificamos que os operadores do conjunto $\{J_{a}^{2},J_{b}^{2},J^{2},J_{3}\}$ também são mutuamente compatíveis.

Vamos denotar os autovetores comuns dos operadores $\{J_{a}^{2},J_{3}^{a},J_{b}^{2},J_{3}^{b}\}$ por $|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle$ e para os autovetores comuns dos operadores $\{J_{a}^{2},J_{b}^{2},J^{2},J_{3}\}$ usaremos $|j_{a},j_{b},j,m\rangle$. Assim, usando $(A\otimes B)(C\otimes D)=AC\otimes BD$, teremos que
\begin{align}
(J_{a}^{2}\otimes\mathbb{I}_{b})(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle) & = (J_{a}^{2}|j_{a},m_{a}\rangle)\otimes|j_{b},m_{b}\rangle \\
& = \hbar^{2}j_{a}(j_{a}+1)|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle, \\
(J_{3}^{a}\otimes\mathbb{I}_{b})(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle) & = (J_{3}^{a}|j_{a},m_{a}\rangle)\otimes|j_{b},m_{b}\rangle \\
& = \hbar m_{a}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle, \\
(\mathbb{I}_{a}\otimes J_{b}^{2})(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle) & = |j_{a},m_{a}\rangle\otimes(J_{b}^{2}|j_{b},m_{b}\rangle) \\
& = \hbar^{2}j_{b}(j_{b}+1)|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle, \\
(\mathbb{I}_{a}\otimes J_{3}^{b})(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle) & = |j_{a},m_{a}\rangle\otimes(J_{3}^{b}|j_{b},m_{b}\rangle) \\
& = \hbar m_{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle
\end{align}
e
\begin{align}
& J_{a}^{2}\otimes\mathbb{I}_{b}|j_{a},j_{b},j,m\rangle = \hbar^{2}j_{a}(j_{a}+1)|j_{a},j_{b},j,m\rangle, \\
& \mathbb{I}_{a}\otimes J_{b}^{2}|j_{a},j_{b},j,m\rangle = \hbar^{2}j_{b}(j_{b}+1)|j_{a},j_{b},j,m\rangle, \\
& J^{2}|j_{a},j_{b},j,m\rangle = \hbar^{2}j(j+1)|j_{a},j_{b},j,m\rangle, \\
& J_{3}|j_{a},j_{b},j,m\rangle = \hbar m|j_{a},j_{b},j,m\rangle.
\end{align}

Notemos ainda que
\begin{align}
J_{3}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle & = (J_{3}^{a}\otimes\mathbb{I}_{b})(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle) + (\mathbb{I}_{a}\otimes J_{3}^{b})(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle) \\
               & = \hbar m_{a}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle + \hbar m_{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle \\
               & = \hbar (m_{a}+m_{b})|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle.
\end{align}

Ou seja, $|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle$ também é autovetor de $J_{3}$ com autovalor $\hbar(m_{a}+m_{b})$. No entanto, __as componentes dos momentos angulares individuias não comutam com o momento angular total__. Por exemplo:
\begin{align}
[J^{2},J_{3}^{a}] & = \sum_{j=1}^{3}[J_{j}^{2},J_{3}^{a}\otimes\mathbb{I}_{b}] \\
& = \sum_{j=1}^{3}\left(J_{j}[J_{j},J_{3}^{a}\otimes\mathbb{I}_{b}] + [J_{j},J_{3}^{a}\otimes\mathbb{I}_{b}]J_{j} \right) \\
                  & = \sum_{j=1}^{3}\left(J_{j}[J_{j}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{j}^{b},J_{3}^{a}\otimes\mathbb{I}_{b}] + [J_{j}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{j}^{b},J_{3}^{a}\otimes\mathbb{I}_{b}]J_{j} \right) \\
                & = \sum_{j=1}^{3}\left(J_{j}([J_{j}^{a},J_{3}^{a}]\otimes\mathbb{I}_{b}+0) + ([J_{j}^{a},J_{3}^{a}]\otimes\mathbb{I}_{b}+0)J_{j} \right) \\
                & = \sum_{j=1}^{3}i\hbar \varepsilon_{j3k}\left(J_{j}J_{k}^{a}\otimes\mathbb{I}_{b} + J_{k}^{a}\otimes\mathbb{I}_{b}J_{j} \right) \\
                & = i\hbar \varepsilon_{13k}\left(J_{1}J_{k}^{a}\otimes\mathbb{I}_{b} + J_{k}^{a}\otimes\mathbb{I}_{b}J_{1} \right) + i\hbar \varepsilon_{23k}\left(J_{2}J_{k}^{a}\otimes\mathbb{I}_{b} + J_{k}^{a}\otimes\mathbb{I}_{b}J_{2} \right) \\
                & = i\hbar \varepsilon_{132}\left(J_{1}J_{2}^{a}\otimes\mathbb{I}_{b} + J_{2}^{a}\otimes\mathbb{I}_{b}J_{1} \right) + i\hbar \varepsilon_{231}\left(J_{2}J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes\mathbb{I}_{b}J_{2} \right) \\
                & = i\hbar\left(J_{2}J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes\mathbb{I}_{b}J_{2} - J_{1}J_{2}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes\mathbb{I}_{b}J_{1} \right) \\
                & = i\hbar(J_{2}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{2}^{b})J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes\mathbb{I}_{b}(J_{2}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{2}^{b}) \\
                & \hspace{1cm} - (J_{1}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{1}^{b})J_{2}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes\mathbb{I}_{b}(J_{1}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{1}^{b})  \\
                & = i\hbar(J_{2}^{a}J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes J_{2}^{b} + J_{1}^{a}J_{2}^{a}\otimes\mathbb{I}_{b}  + J_{1}^{a}\otimes J_{2}^{b} \\
                & \hspace{1cm} - J_{1}^{a}J_{2}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes J_{1}^{b} - J_{2}^{a}J_{1}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes J_{1}^{b}) \\
                & = 2i\hbar(J_{1}^{a}\otimes J_{2}^{b} - J_{2}^{a}\otimes J_{1}^{b}) \\
                & \ne 0.                
\end{align}
Analogamente, pode-se verificar que
\begin{equation}
[J^{2},J_{3}^{b}] \ne 0.
\end{equation}

Estes últimos resultados nos mostram que, para os momentos angulares individuais fixos (i.e., para os valores dos números quânticos $j_{s}$ fixos), estados $|j_{a},j_{b},j,m\rangle$ com momento angular total fixo devem corresponder a superposições de dois ou mais estados $|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle$ com valores fixos das componentes $z$ dos momentos angulares individuais. Para escrevermos essa superposição explicitamente, vamos utilizar a relação de completeza:
\begin{equation}
\sum_{m_{s}=-j_{s}}^{j_{s}}|j_{s},m_{s}\rangle\langle j_{s},m_{s}| = \mathbb{I}_{j_{s}}
\end{equation}
para escrever o operador identidade no subespaço $\mathcal{H}_{j_{a}}\otimes\mathcal{H}_{j_{b}}\subset\mathcal{H}$ gerado por estados com $\{j_{s}\}_{s=a,b}$ constantes. Segue assim que
\begin{align}
|j_{a},j_{b},j,m\rangle & = \mathbb{I}_{j_{a}}\otimes\mathbb{I}_{j_{b}}|j_{a},j_{b},j,m\rangle \\
                        & = \left(\sum_{m_{a}=-j_{a}}^{j_{a}}|j_{a},m_{a}\rangle\langle j_{a},m_{a}|\otimes\sum_{m_{b}=-j_{b}}^{j_{b}}|j_{b},m_{b}\rangle\langle j_{b},m_{b}|\right)|j_{a},j_{b},j,m\rangle \\
                        & = \sum_{m_{a}=-j_{a}}^{j_{a}}\sum_{m_{b}=-j_{b}}^{j_{b}}(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle)(\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)|j_{a},j_{b},j,m\rangle \\
                        & =: \sum_{m_{a}=-j_{a}}^{j_{a}}\sum_{m_{b}=-j_{b}}^{j_{b}} C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle,
\end{align}
onde usamos novamente $AB\otimes CD=(A\otimes C)(B\otimes D)$ e definimos os chamados __coeficientes de Clebsch-Gordan__ como:
\begin{equation}
C_{m_{a}m_{b}m}^{j_{a}j_{b}j} := (\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)|j_{a},j_{b},j,m\rangle.
\end{equation}

Existem tabelas desses coeficientes, cuja obtenção usa restrições algébricas e simetrias. Por exemplo, uma condição necessária para que os coeficientes sejam não nulos é que
\begin{equation}
m=m_{a}+m_{b}.
\end{equation}
Para verificar essa afirmação consideremos ("atua" $J_{3}$ na esquerda e $J_{3}^{s}$ na direita)
\begin{align}
& (\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)(J_{3} - J_{3}^{a}\otimes\mathbb{I}_{b} - \mathbb{I}_{a}\otimes J_{3}^{b})|j_{a},j_{b},j,m\rangle = (\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)\mathbb{O}|j_{a},j_{b},j,m\rangle = 0, \\
& \therefore \hbar(m-m_{a}-m_{b})(\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)|j_{a},j_{b},j,m\rangle = 0, \\
& \therefore \hbar(m-m_{a}-m_{b})C_{m_{a}m_{b}m}^{j_{a}j_{b}j} = 0.
\end{align}

OBS. Como a base $|j_{s},m_{s}\rangle$ possiu $2j_{s}+1$ vetores, temos a dimensão
\begin{equation}
\dim(\mathcal{H}_{j_{a}}\otimes\mathcal{H}_{j_{b}}) = (2j_{a}+1)(2j_{b}+1),
\end{equation}
que é igual ao número de estados da base $\{|j_{a},j_{b},j,m\rangle\}$. 

## Possíveis valores de $j$ pra $j_s$ fixos
Agora, fixados os valores de $j_{s}$, se conhecermos também os possíveis valores de $m_{s}$, e consequentemente os possíveis valores de $m$, para obter os possíveis valores de $j$, notemos que o maior momento angular total é obtido se $\vec{J}_{a}$ e $\vec{J}_{b}$ forem "paralelos", ou seja, se tiverem ambos e.g. os valores máximos das componentes $z$ dos momentos angulares individuais, i.e., se $\{m_{s}=j_{s}\}_{s=a,b}$. Assim
\begin{equation}
m=m_{a}+m_{b}=j_{a}+j_{b}=j_{\max}.
\end{equation}
Por outro lado, o menor valor possível de $j$ é obtido se $\vec{J}_{a}$ e $\vec{J}_{b}$ forem "antiparalelos", ou seja, se e.g. $m_{a}=j_{a}$ e $m_{b}=-j_{b}$ ou $m_{a}=-j_{a}$ e $m_{b}=j_{b}$. Para ter o mínimo entre $m=j_{a}-j_{b}$ e $m=j_{b}-j_{a}$ usamos como o menor valor de $j$ o módulo $|j_{a}-j_{b}|$, i.e., 
\begin{equation}
j_{min}=\min(j_{a}-j_{b},j_{b}-j_{a}) = |j_{a}-j_{b}|.
\end{equation}
Com isso segue que $j\in\mathbb{N}$ com
\begin{equation}
j=|j_{a}-j_{b}|,|j_{a}-j_{b}|+1,\cdots,j_{a}+j_{b}.
\end{equation}

## __Exemplo:__ Soma de dois spins 1/2
Como um primeiro exemplo, consideremos que $j_{s}=1/2$ para $s=a,b$. Nesse caso 
$$m_{s}=-1/2,1/2,\ j=0,1,\ m(j=0)=0 \text{ e } m(j=1)=-1,0,1.$$
Vamos determinar os coeficientes de Clebsch-Gordan nesse caso. Por conveniência, não escreveremos $j_{s}$ nos labels dos vetores de estado, ou seja, denotaremos
$$|j_s,m_s\rangle\equiv|m_s\rangle \text{ e } |j_a,j_b,j,m\rangle\equiv|j,m\rangle.$$
Usando a restrição 
$$m=m_{a}+m_{b}$$
e começando com o estado de maior componente $z$ do momento angular total, vemos que
\begin{align}
|j=1,m=1\rangle & = \sum_{m_{a}=-1/2}^{1/2}\sum_{m_{b}=-1/2}^{1/2}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle \\
                & = C_{-1/2,-1/2,1}^{1/2,1/2,1}|-1/2\rangle\otimes|-1/2\rangle + C_{-1/2,1/2,1}^{1/2,1/2,1}|-1/2\rangle\otimes|1/2\rangle + C_{1/2,-1/2,1}^{1/2,1/2,1}|1/2\rangle\otimes|-1/2\rangle + C_{1/2,1/2,1}^{1/2,1/2,1}|1/2\rangle\otimes|1/2\rangle \\
                & = 0 + 0 + 0 + C_{1/2,1/2,1}^{1/2,1/2,1}|1/2\rangle\otimes|1/2\rangle \\
                & = |m_{a}=1/2\rangle\otimes|m_{b}=1/2\rangle.
\end{align}

Seguindo, aplicamos o operador de abaixamento $J_{-} = J_{1}-iJ_{2}$ nesse estado, lembrando que $J_{-}|j,m\rangle=\hbar\sqrt{j(j+1)-m(m-1)}|j,m-1\rangle$,
para obter $J_{-}|j=1,m=1\rangle=\hbar\sqrt{1(1+1)-1(1-1)}|1,1-1\rangle=\hbar\sqrt{2}|1,0\rangle$. Assim
\begin{align}
|j=1,m=0\rangle & = (1/\hbar\sqrt{2})J_{-}|j=1,m=1\rangle \\
                & = (1/\hbar\sqrt{2})(J_{-}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{-}^{b})|1/2\rangle\otimes|1/2\rangle \\
                & = (1/\hbar\sqrt{2})((J_{-}^{a}|1/2\rangle)\otimes|1/2\rangle + |1/2\rangle\otimes(J_{-}^{b}|1/2\rangle)) \\
                & = (1/\hbar\sqrt{2})((\hbar\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}-1)}|1/2-1\rangle)\otimes|1/2\rangle + |1/2\rangle\otimes(\hbar\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}-1)}|1/2-1\rangle)) \\
                & = (1/\sqrt{2})(\sqrt{1/2(1/2+1)-(1/2)(1/2-1)}|-1/2\rangle\otimes|1/2\rangle + \sqrt{(1/2)(1/2+1)-(1/2)(1/2-1)}|1/2\rangle\otimes|-1/2\rangle) \\
                & = (1/\sqrt{2})(\sqrt{3/4+1/4}|-1/2\rangle\otimes|1/2\rangle + \sqrt{3/4+1/4}|1/2\rangle\otimes|-1/2\rangle) \\
                & = (1/\sqrt{2})|m_{a}=-1/2\rangle\otimes|m_{b}=1/2\rangle + (1/\sqrt{2})|m_{a}=1/2\rangle\otimes|m_{b}=-1/2\rangle.
\end{align}

Ainda, $J_{-}|j=1,m=0\rangle=\hbar\sqrt{1(1+1)-0(0-1)}|1,0-1\rangle=\hbar\sqrt{2}|j=1,m=-1\rangle$. Assim
\begin{align}
|j=1,m=-1\rangle & = (\hbar^{-1}/\sqrt{2})J_{-}|j=1,m=0\rangle \\
                 & = (\hbar^{-1}/2)(J_{-}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{-}^{b})(|-1/2\rangle\otimes|1/2\rangle + |1/2\rangle\otimes|-1/2\rangle) \\
                 & = (\hbar^{-1}/2)(J_{-}^{a}\otimes\mathbb{I}_{b}|-1/2\rangle\otimes|1/2\rangle + J_{-}^{a}\otimes\mathbb{I}_{b}|1/2\rangle\otimes|-1/2\rangle + \mathbb{I}_{a}\otimes J_{-}^{b}|-1/2\rangle\otimes|1/2\rangle + \mathbb{I}_{a}\otimes J_{-}^{b}|1/2\rangle\otimes|-1/2\rangle) \\
                 & = (1/2)(|\oslash\rangle\otimes|1/2\rangle + \sqrt{1}|-1/2\rangle\otimes|-1/2\rangle + \sqrt{1}|-1/2\rangle\otimes|-1/2\rangle + |1/2\rangle\otimes|\oslash\rangle) \\
                 & = |m_{a}=-1/2\rangle\otimes|m_{b}=-1/2\rangle.
\end{align}

Construíremos o estado que falta da base como o primeiro
\begin{align}
|j=0,m=0\rangle & = \sum_{m_{a}=-1/2}^{1/2}\sum_{m_{b}=-1/2}^{1/2}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle \\
                & = C_{-1/2,-1/2,0}^{1/2,1/2,0}|-1/2\rangle\otimes|-1/2\rangle + C_{-1/2,1/2,0}^{1/2,1/2,0}|-1/2\rangle\otimes|1/2\rangle + C_{1/2,-1/2,0}^{1/2,1/2,0}|1/2\rangle\otimes|-1/2\rangle + C_{1/2,1/2,0}^{1/2,1/2,0}|1/2\rangle\otimes|1/2\rangle \\
                & = 0 + C_{-1/2,1/2,0}^{1/2,1/2,0}|-1/2\rangle\otimes|1/2\rangle + C_{1/2,-1/2,0}^{1/2,1/2,0}|1/2\rangle\otimes|-1/2\rangle + 0 \\
                & = (1/\sqrt{2})\left(|m_{a}=-1/2\rangle\otimes|m_{b}=1/2\rangle - |m_{a}=1/2\rangle\otimes|m_{b}=-1/2\rangle\right),
\end{align}
com a escolha dos coeficientes de Clebsch-Gordan feita de forma que o estado seja normalizado e ortogonal aos outros estados da base, i.e.,
\begin{equation}
|||j=0,m=0\rangle||=1 \therefore |C_{-1/2,1/2,0}^{1/2,1/2,0}|^{2}+|C_{1/2,-1/2,0}^{1/2,1/2,0}|^{2}=1
\end{equation}
e
\begin{equation}
\langle j=1,m=0|j=0,m=0\rangle=0 \therefore (1/\sqrt{2})C_{-1/2,1/2,0}^{1/2,1/2,0}+(1/\sqrt{2})C_{1/2,-1/2,0}^{1/2,1/2,0}=0.
\end{equation}

### Relação de recursão para os coeficientes de Clebsch-Gordan
Vamos considerar $j_{a},j_{b},j$ fixos. De $|j,m\rangle=\sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle$ teremos que
\begin{align}
J_{\pm}|j,m\rangle & = \hbar\sqrt{j(j+1)-m(m\pm1)}|j,m\pm1\rangle \\
        & = (J_{\pm}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{\pm}^{b})|j,m\rangle \\
        & = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\left(J_{\pm}^{a}|m_{a}\rangle\otimes|m_{b}\rangle + |m_{a}\rangle\otimes J_{\pm}^{b}|m_{b}\rangle\right) \\
        & = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\hbar\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\pm1)}|m_{a}\pm1\rangle\otimes|m_{b}\rangle \\
         & + \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\hbar\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\pm1)}|m_{a}\rangle\otimes|m_{b}\pm1\rangle. 
\end{align}
Tomando o produto interno de $|m_{a}'\rangle\otimes|m_{b}'\rangle$ com esta última equação dividida por $\hbar$ vem que
\begin{align}
& \sqrt{j(j+1)-m(m\pm1)}\langle m_{a}'|\otimes\langle m_{b}'|j,m\pm1\rangle \\
& = \sqrt{j(j+1)-m(m\pm1)}C_{m_{a}'m_{b}'(m\pm1)}^{j_{a}j_{b}j} \\
& = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\pm1)}(\langle m_{a}'|\otimes\langle m_{b}'|)(|m_{a}\pm1\rangle\otimes|m_{b}\rangle) \\ 
& + \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\pm1)}(\langle m_{a}'|\otimes\langle m_{b}'|)(|m_{a}\rangle\otimes|m_{b}\pm1\rangle) \\
& = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\pm1)}\delta_{m_{a}',m_{a}\pm1}\delta_{m_{b}',m_{b}} \\ 
& + \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\pm1)}\delta_{m_{a}',m_{a}}\delta_{m_{b}',m_{b}\pm1} \\
& = C_{(m_{a}'\mp1)m_{b}'m}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}'(m_{a}'\mp1)} \\
& + C_{m_{a}'(m_{b}'\mp1)m}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}'(m_{b}'\mp1)}.
\end{align}
Usando $m'=m\pm1$ e trocando os números quânticos com linha pelos sem obtemos
\begin{align}
\sqrt{j(j+1)-m(m\mp1)}C_{m_{a}m_{b}m}^{j_{a}j_{b}j} & = C_{(m_{a}\mp1)m_{b}(m\mp1)}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\mp1)} \\
& + C_{m_{a}(m_{b}\mp1)(m\mp1)}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\mp1)}.
\end{align}

```python
def cJp(j,m):
    hb = symbols('hbar')
    return hb*sqrt(j*(j+1)-m*(m+1))
def cJm(j,m):
    hb = symbols('hbar')
    return hb*sqrt(j*(j+1)-m*(m-1))
```

```python
def clebsch_gordan(ja,jb):
    nma = int(2*ja+1)
    nmb = int(2*jb+1)
    nj = int(ja+jb-abs(ja-jb)+1)
    print(' ja = ',ja,' jb = ',jb)
    for k in range(0,nj):
        j = ja+jb-k
        nm = int(2*j+1)
        m = j
        print('j=',j,'m=',m)
        print('ma=',ja,', mb=',jb,', ccg=',1,', Os outros coeficientes são nulos')
        for l in range(1,nm):
            m = j-l
            print('j = ',j,', m = ',m)
            norm = cJm(j,m)
            for p in range(0,nma):
                ma = ja-p
                for q in range(0,nmb):
                    mb = jb-q
                    if m != (ma+mb):
                        ccg = 0
                    else:
                        if ma != (-ja):
                            ccg = cJm(ja,ma)/norm 
                        else:
                            ccg = cJm(jb,mb)/norm 
                    print('ma = ',ma,', mb = ',mb,', ccg = ',ccg)
                        
    return
clebsch_gordan(1/2,1/2)
```
