# Teoria de perturbação
O objetivo de teoria de perturbação é construir soluções aproximadas (estáticas e/ou dinâmicas) a partir das soluções analíticas obtidas em casos particulares, que geralmente estão relacionados com intensidades de interação maiores em relação às correções que queremos incluir.

## Teoria de perturbação independente do tempo
### Espectros discretos não degenerados
Para Hamiltonianos com espectro discreto não degenerado, começamos considerando que nosso objetivo é obter a solução da equação de autovalores e autovetores:
\begin{equation}
H|\psi_{n}\rangle = E_{n}|\psi_{n}\rangle,
\end{equation}
quando o Hamiltoniano pode ser escrito como 
\begin{equation}
H = H_{0} + H_{1},
\end{equation}
onde $H_{0}$ representa a parte mais intensa da interação e cujos autovalores e autovetores são conhecidos, i.e., conhecemos a solução de 
\begin{equation}
H_{0}|\psi_{n}^{(0)}\rangle = E_{n}^{(0)}|\psi_{n}^{(0)}\rangle,
\end{equation}
e $H_{1}$ é a perturbação, uma interação de intensidade pequena em relação a $H_{0}$. Na sequência obteremos equações que nos possibilitarão, a princípio, calcular correções parturbativas de várias ordens para a energia. Começaremos trocando $H = H_{0} + H_{1}$ por
\begin{equation}
H = H_{0} + \xi H_{1},
\end{equation}
e depois fazemos $\xi=1$. Seguindo, escrevemos a solução geral em termos da não perturbada como segue:
\begin{equation}
|\psi_{n}\rangle = N_{n}(|\psi_{n}^{(0)}\rangle + |\chi_{n}\rangle),
\end{equation}
com $|\chi_{n}\rangle$ não sendo necessariamente normalizado e satisfazendo $\langle\chi_{n}|\psi_{n}^{(0)}\rangle=0$. <br>
__Exercício:__ Verifique que a constate de normalização, considerada $\in\mathbb{R}$, é dada por
\begin{equation}
N_{n} = \frac{1}{\sqrt{1+\langle\chi_{n}|\chi_{n}\rangle}}.
\end{equation}

Subistituindo o Hamiltoniano e autovetors como escritos acima, teremos
\begin{align}
& H|\psi_{n}\rangle = (H_{0}+\xi H_{1})N_{n}(|\psi_{n}^{(0)}\rangle + |\chi_{n}\rangle ) \\
& = N_{n}\left(H_{0}|\psi_{n}^{(0)}\rangle + H_{0}|\chi_{n}\rangle + \xi H_{1}|\psi_{n}^{(0)}\rangle + \xi H_{1}|\chi_{n}\rangle\right) \\
& = N_{n}\left(E_{n}^{(0)}|\psi_{n}^{(0)}\rangle + H_{0}|\chi_{n}\rangle + \xi H_{1}|\psi_{n}^{(0)}\rangle + \xi H_{1}|\chi_{n}\rangle\right) \\
& = E_{n}|\psi_{n}\rangle \\
& = E_{n}N_{n}(|\psi_{n}^{(0)}\rangle + |\chi_{n}\rangle).
\end{align}
Desta equação, prontamente identificamos a relação:
\begin{equation}
(E_{n}-E_{n}^{(0)})|\psi_{n}^{(0)}\rangle = \xi H_{1}|\psi_{n}^{(0)}\rangle + (H_{0}+ \xi H_{1} - E_{n})|\chi_{n}\rangle. \text{ (*)}
\end{equation}
Se tomarmos o produto interno de $|\psi_{n}^{(0)}\rangle$ com ($*$):
\begin{align}
& (E_{n}-E_{n}^{(0)})\langle\psi_{n}^{(0)}|\psi_{n}^{(0)}\rangle = E_{n}-E_{n}^{(0)} \\
& = \xi \langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \langle\psi_{n}^{(0)}|H_{0}|\chi_{n}\rangle + \xi \langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle - E_{n}\langle\psi_{n}^{(0)}|\chi_{n}\rangle \\
& = \xi \langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + E_{n}^{(0)}\langle\psi_{n}^{(0)}|\chi_{n}\rangle + \xi \langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle - 0 \\
& = \xi\left(\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle\right).
\end{align}
ou
\begin{equation}
E_{n} = E_{n}^{(0)} + \xi\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \xi\langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle.  \text{ (**)}
\end{equation}

Note que na última expressão somente último termo está escrito em termos de $|\chi_{n}\rangle$, que não conhecemos. Vamos então decompor este vetor na base $\{|\psi_{n}^{(0)}\rangle\}$:
\begin{align}
|\chi_{n}\rangle & = \sum_{m}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|\chi_{n}\rangle \\
                 & = \sum_{m\ne n}\langle\psi_{m}^{(0)}|\chi_{n}\rangle|\psi_{m}^{(0)}\rangle.
\end{align}
Para obter os coeficientes $\langle\psi_{m}^{(0)}|\chi_{n}\rangle$ para $m\ne n$ tomamos o produto interno de $|\psi_{m}^{(0)}\rangle$ com ($*$):
\begin{align}
& (E_{n}-E_{n}^{(0)})\langle\psi_{m}^{(0)}|\psi_{n}^{(0)}\rangle = 0 \\
& = \xi \langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \langle\psi_{m}^{(0)}|H_{0}|\chi_{n}\rangle + \xi \langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle - E_{n}\langle\psi_{m}^{(0)}|\chi_{n}\rangle \\
& = \xi \langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + E_{m}^{(0)}\langle\psi_{m}^{(0)}|\chi_{n}\rangle + \xi \langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle - E_{n}\langle\psi_{m}^{(0)}|\chi_{n}\rangle.
\end{align}
Assim, como $E_{n}\ne E_{m}^{(0)}$,
\begin{equation}
\langle\psi_{m}^{(0)}|\chi_{n}\rangle = \xi\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}} + \xi\frac{\langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle}{E_{n}-E_{m}^{(0)}}.
\end{equation}
Usando esses coeficientes na decomposição de $|\chi_{n}\rangle$ e o substituindo iterativamente vem que
\begin{align}
|\chi_{n}\rangle & = \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle + \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle \\
& = \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle \\ 
& + \xi^{2}\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}\frac{\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m'}^{(0)}}|\psi_{m}^{(0)}\rangle \\
& + \cdots.
\end{align}

Substituindo essa equação em $(**)$ obteremos
\begin{align}
E_{n} & = E_{n}^{(0)} + \xi\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle \\
& + \xi^{2}\sum_{m\ne n}\frac{\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}} \\
& + \xi^{3}\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{(E_{n}-E_{m}^{(0)})(E_{n}-E_{m'}^{(0)})} \\
& + \cdots.
\end{align}

Agora vamos utilizar a dita _expansão de Rayleigh-Schrödinger_. Notemos que ainda nos resta o $E_{n}$ no denominador das expansões acima. Identificando
\begin{equation}
E_{n} = \sum_{j=0}^{\infty} E_{n}^{(j)}\xi^{j}
\end{equation}
e usando a série $(1-x)^{-1} = \sum_{j=0}^{\infty}x^{j}$ teremos que
\begin{align}
\frac{1}{E_{n}-E_{m}^{(0)}} & = \frac{1}{E_{n}^{(0)}+\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}-E_{m}^{(0)}} \\
& = \frac{1}{(E_{n}^{(0)}-E_{m}^{(0)})}\frac{1}{(1+\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}/(E_{n}^{(0)}-E_{m}^{(0)}))} \\
& = \frac{1}{(E_{n}^{(0)}-E_{m}^{(0)})}\left(1 - \frac{\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}}{E_{n}^{(0)}-E_{m}^{(0)}} + \frac{(\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j})^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} - \cdots\right) \\
& = \frac{1}{E_{n}^{(0)}-E_{m}^{(0)}} - \frac{\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} + \frac{(\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j})^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{3}} - \cdots \\
& = \frac{1}{E_{n}^{(0)} - E_{m}^{(0)}} - \xi\frac{E_{n}^{(1)}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} \\
& + \xi^{2}\left( \frac{E_{n}^{(2)}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} + \frac{(E_{n}^{(1)})^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{3}}\right) - \cdots.
\end{align}

Se substituímos esta expressão naquela para $E_{n}$ acima obteremos as várias _ordens de correção para a energia_:
\begin{align}
E_{n} & = \xi^{0}E_{n}^{(0)} + \xi E_{n}^{(1)} + \xi^{2}E_{n}^{(2)} + \xi^{3}E_{n}^{(3)} + \cdots \\
& = \xi^{0}E_{n}^{(0)} + \xi\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \xi^{2}\sum_{m\ne n}\frac{|\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle|^{2}}{E_{n}^{(0)}-E_{m}^{(0)}} \\
& + \xi^{3}\left(\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{(E_{n}^{(0)}-E_{m}^{(0)})(E_{n}^{(0)}-E_{m'}^{(0)})} - E_{n}^{(1)}\sum_{m\ne n}\frac{|\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle|^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} \right) \\
& + \cdots.
\end{align}

Fazendo o análogo para $|\chi_{n}\rangle$ obteremos
\begin{align}
|\chi_{n}\rangle & = \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}^{(0)}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle \\ 
& + \xi^{2}\left(\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle}{E_{n}^{(0)}-E_{m}^{(0)}}\frac{\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}^{(0)}-E_{m'}^{(0)}}|\psi_{m}^{(0)}\rangle - E_{n}^{(1)}\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}}|\psi_{m}^{(0)}\rangle \right) \\
& + \cdots.
\end{align}
_OBS:_ Lembre que antes de utilizar os autoestados $|\psi_{n}\rangle = N_{n}(|\psi_{n}^{(0)}\rangle+|\chi_{n}\rangle)$ para calcular probabilidades, você deve obter a constante de normalização.

### Estrutura fina do átomo de Hidrogênio
Lembrando, com o referencial no centro de massa, o Hamiltoniano do átomo de Hidrogênio tem a forma
\begin{equation}
H_{0} = \frac{P^{2}}{2m}+V = -\frac{\hbar^{2}}{2m}\nabla^{2} - \frac{e^{2}}{4\pi\epsilon_{0}}\frac{1}{r},
\end{equation}
onde $m=m_{e}m_{p}/(m_{e}+m_{p})$. Da solução exata para a equação de Schrödinger temos as energias
\begin{equation}
E_{n}^{(0)} = -\left[\frac{m}{2\hbar^{2}}\left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\right]\frac{1}{n^{2}} = \frac{-13,6\text{ eV}}{n^{2}},
\end{equation}
com $n=1,2,3,\cdots$. Também utilizaremos o raio de Bohr
\begin{equation}
a = \frac{4\pi\epsilon_{0}\hbar^{2}}{me^{2}} \approx 0,528\text{ }\mathring{A}
\end{equation}
e a constante de estrutura fina
\begin{equation}
\alpha = \frac{e^{2}}{4\pi\epsilon_{0}\hbar c} \approx \frac{1}{137,036}.
\end{equation}

#### Correção relativísitica
Lembrando, a energia relativística é dada por
\begin{equation}
T = (\gamma-1)mc^{2},
\end{equation}
com $\gamma = 1/\sqrt{1-v^{2}/c^{2}}$. Ademais, o momento linear relativístico é $P=\gamma mv$. <br>
__Exercício:__ Verifique que
\begin{align}
T & = \sqrt{P^{2}c^{2}+m^{2}c^{4}}-mc^{2} \\ 
  & = mc^{2}\left(\sqrt{1+P^{2}/m^{2}c^{2}} - 1\right).
\end{align}
Seguindo usamos a série $\sqrt{1+x}=1+x/2-x^{2}/8+x^{3}/16-\cdots$ para escrever
\begin{align}
T & = mc^{2}\left(1 + \frac{P^{2}}{2m^{2}c^{2}} - \frac{P^{4}}{8m^{4}c^{4}} + \cdots - 1\right) \\
  & \approx \frac{P^{2}}{2m} - \frac{P^{4}}{8m^{3}c^{2}}.
\end{align}
Então, escrevendo o Hamiltoniano total, nesta aproximação, como $H=H_{0}+H_{1}^{r}$, a correção relativísitca será calculada usando
\begin{equation}
H_{1}^{r}=-\frac{P^{4}}{8m^{3}c^{2}}.
\end{equation}

Correção perturbativa de primeira ordem:
\begin{align}
E_{n}^{(r1)} & = \langle\psi_{n,l,m_{l}}^{(0)}|H_{1}^{r}|\psi_{n,l,m_{l}}^{(0)}\rangle \\
            & = -\frac{1}{8m^{3}c^{2}}\langle\psi_{n,l,m_{l}}^{(0)}|P^{4}|\psi_{n,l,m_{l}}^{(0)}\rangle.
\end{align}
O termo $P^{4}$ pode ser reescrito como segue
\begin{align}
P^{4} & = P^{2}P^{2} \\
      & = 2m(H_{0}-V)2m(H_{0}-V) \\
      & = 4m^{2}\left(H_{0}^{2} - VH_{0} - H_{0}V + V^{2}\right).
\end{align}
Assim
\begin{align}
E_{n}^{(r1)} & = -\frac{1}{2mc^{2}}\left(\langle \psi_{n,l,m_{l}}^{(0)}|H_{0}^{2}|\psi_{n,l,m_{l}}^{(0)}\rangle - \langle\psi_{n,l,m_{l}}^{(0)}|VH_{0}|\psi_{n,l,m_{l}}^{(0)}\rangle - \langle\psi_{n,l,m_{l}}^{(0)}|H_{0}V|\psi_{n,l,m_{l}}^{(0)}\rangle + \langle\psi_{n,l,m_{l}}^{(0)}|V^{2}|\psi_{n,l,m_{l}}^{(0)}\rangle \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} -2E_{n}^{(0)}\langle\psi_{n,l,m_{l}}^{(0)}|V|\psi_{n,l,m_{l}}^{(0)}\rangle + \langle\psi_{n,l,m_{l}}^{(0)}|V^{2}|\psi_{n,l,m_{l}}^{(0)}\rangle \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} +2E_{n}^{(0)}\frac{e^{2}}{4\pi\epsilon_{0}}\langle\psi_{n,l,m_{l}}^{(0)}|r^{-1}|\psi_{n,l,m_{l}}^{(0)}\rangle + \left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\langle\psi_{n,l,m_{l}}^{(0)}|r^{-2}|\psi_{n,l,m_{l}}^{(0)}\rangle \right).
\end{align}
Usando (__exercício__) $\langle\psi_{n,l,m_{l}}^{(0)}|r^{-1}|\psi_{n,l,m_{l}}^{(0)}\rangle=1/an^{2}$ e $\langle\psi_{n,l,m_{l}}^{(0)}|r^{-2}|\psi_{n,l,m_{l}}^{(0)}\rangle=1/a^{2}n^{3}(l+1/2)$ teremos
\begin{align}
E_{n}^{(r1)} & = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} + 2E_{n}^{(0)}\frac{e^{2}}{4\pi\epsilon_{0}}\frac{1}{an^{2}} + \left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\frac{1}{a^{2}n^{3}(l+1/2)} \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} + 2E_{n}^{(0)}\frac{e^{2}}{4\pi\epsilon_{0}}\frac{me^{2}}{4\pi\epsilon_{0}\hbar^{2}}\frac{1}{n^{2}} + \left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\frac{m^{2}e^{4}}{(4\pi\epsilon_{0}\hbar^{2})^{2}}\frac{1}{n^{3}(l+1/2)} \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} - 4(E_{n}^{(0)})^{2} + (E_{n}^{(0)})^{2}\frac{4n}{l+1/2} \right) \\
& = -\frac{1}{2mc^{2}}(E_{n}^{(0)})^{2}\left(1 - 4 + \frac{4n}{l+1/2} \right) \\
& = -\frac{(E_{n}^{(0)})^{2}}{2mc^{2}}\left(\frac{4n}{l+1/2} -3 \right).
\end{align}

Com a correção relativística, as energias do átomo de Hidrogênio ficam
\begin{equation}
E_{n} \rightarrow E_{n}^{(0)}\left(1 - \frac{E_{n}^{(0)}}{2mc^{2}}\left(\frac{4n}{l+1/2} -3 \right)\right).
\end{equation}
##### _Ordem de magnitude da correção relativística_:
Considera o caso $n=1$ e $l=0$, para o qual
\begin{align}
E_{1}^{(r1)} & = -\frac{(-13,6\text{ eV})^{2}}{2\mathrm{x}5,11\mathrm{x}10^{5}\text{ eV}}\left(\frac{4\mathrm{x}1}{0+1/2} -3 \right) \\ 
             & = 9,05\mathrm{x}10^{-4}\text{ eV}, 
\end{align}
onde usamos
\begin{align}
mc^{2} & = \frac{m_{e}m_{p}c^{2}}{m_{e}+m_{p}} \\
& = \frac{(9,11\mathrm{x}10^{-31}\text{ kg})(1,67\mathrm{x}10^{-27}\text{ kg})(3\mathrm{x}10^{8}\text{ m/s})}{9,11\mathrm{x}10^{-31}\text{ kg}+1,67\mathrm{x}10^{-27}\text{ kg}} \\
& = 8,18\mathrm{x}10^{-14}\text{ J} = 8,18\mathrm{x}10^{-14}\frac{\text{1 eV}}{1,6\mathrm{x}10^{-19}} \\
& = 5,11\mathrm{x}10^{5}\text{ eV}.
\end{align}

#### Interação spin-órbita
Do referencial do elétron, o próton está em movimento e gera um campo magnético que pode ser calculado usando a lei de Biot-Savart:
\begin{align}
\vec{B} & = \frac{\mu_{0}i}{4\pi}\oint_{c}\frac{d\vec{l}\mathrm{x}\hat{r}}{r^{2}} \\
& = \frac{\mu_{0}i}{4\pi r^{2}}\oint_{c}ds\hat{e} \\
& = \frac{\mu_{0}i}{4\pi r^{2}}2\pi r\hat{e} \\
& = \frac{\mu_{0}e/\tau}{2r}\hat{e},
\end{align}
em que aproximaos a "trajetória" por um círculo. O momento angular do próton, nessas condições, será
\begin{equation}
L = rmv = rm2\pi r/\tau.
\end{equation}
Assim
\begin{align}
\vec{B} & = \frac{\mu_{0}e}{2r}\frac{L}{m2\pi r^{2}}\hat{e} \\
& = \frac{\mu_{0}e}{4\pi mr^{3}}\vec{L} \\
& = \frac{e}{4\pi\epsilon_{0} mc^{2}r^{3}}\vec{L},
\end{align}
onde usamos $c=1/\sqrt{\mu_{0}\epsilon_{0}}$. Este campo magnético interage como o momento magnético de spin do elétron $\vec{\mu}_{e} = (-e/m)\vec{S}_{e}$ segundo o Hamiltoniano:
\begin{align}
H_{1}^{so} & = -\mu_{e}\cdot \vec{B} \\
& = \frac{e^{2}}{8\pi\epsilon_{0} m^{2}c^{2}r^{3}}\vec{S}\cdot\vec{L},
\end{align}
em que introdizimos ainda o fator 1/2 para a precessão de Thomas para o referencial não incercial do elétron.

A correção perturbativa de primeira ordem devido à interação spin-órbita será dada por
\begin{align}
E_{n}^{(so1)} & = \langle\psi_{n,l,m_{l}}^{(0)}|H_{1}^{so}|\psi_{n,l,m_{l}}^{(0)}\rangle \\
            & = \frac{e^{2}}{8\pi\epsilon_{0} m^{2}c^{2}}\langle\psi_{n,l,m_{l}}^{(0)}|\frac{\vec{S}\cdot\vec{L}}{r^{3}}|\psi_{n,l,m_{l}}^{(0)}\rangle.
\end{align}
Para o momento angular total do elétron,
\begin{equation}
\vec{J} = \mathbb{I}_{L}\otimes\vec{S} + \vec{L}\otimes\mathbb{I}_{S},
\end{equation}
teremos que (__exercício__)
\begin{align}
\vec{S}\cdot\vec{L} & = (\mathbb{I}_{L}\otimes\vec{S})\cdot(\vec{L}\otimes\mathbb{I}_{S}) \\ 
& = 2^{-1}\left(J^{2} - \mathbb{I}_{L}\otimes S^{2} - L^{2}\otimes\mathbb{I}_{S} \right).
\end{align}

Assim, trocando $|\psi_{n,l,m_{l}}^{(0)}\rangle = |n,l,s,j,m_{j}\rangle$ teremos
\begin{align}
E_{n}^{(so1)} & = \frac{e^{2}}{8\pi\epsilon_{0} m^{2}c^{2}}\langle n,l,s,j,m_{j}|\frac{2^{-1}\left(J^{2} - \mathbb{I}_{L}\otimes S^{2} - L^{2}\otimes\mathbb{I}_{S} \right)}{r^{3}}|n,l,s,j,m_{j}\rangle \\
& = \frac{e^{2}\hbar^{2}\left(j(j+1)-l(l+1)-s(s+1)\right)}{16\pi\epsilon_{0} m^{2}c^{2}}\langle n,l,s,j,m_{j}|r^{-3}|n,l,s,j,m_{j}\rangle.
\end{align}
Usando $\langle n,l,s,j,m_{j}|r^{-3}|n,l,s,j,m_{j}\rangle = 1/l(l+1/2)(l+1)n^{3}a^{3}$ e $s=1/2$ vem que
\begin{align}
E_{n}^{(so1)} & = \frac{e^{2}\hbar^{2}\left(j(j+1)-l(l+1)-3/4\right)}{16\pi\epsilon_{0} m^{2}c^{2}}\frac{1}{l(l+1/2)(l+1)n^{3}a^{3}} \\
& = \frac{e^{2}\hbar^{2}}{16\pi\epsilon_{0}m^{2}c^{2}}\frac{1}{n^{3}}\frac{m^{3}e^{6}}{(4\pi\epsilon_{0})^{3}\hbar^{6}}\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)} \\
& = \frac{E_{n}^{(0)})^{2}n}{mc^{2}}\left(\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)}\right).
\end{align}
Vemos assim que a correção devido à interação spin-órbita tem a mesma ordem de magnitude que a correção relativística. Vamos então adicionar essas duas contribuições para obter a estrutura fina do átomo de Hidrogênio:
\begin{align}
E_{n} & \rightarrow E_{n}^{(0)} + E_{n}^{r(0)} + E_{n}^{(so0)} \\
& = E_{n}^{(0)} -\frac{(E_{n}^{(0)})^{2}}{2mc^{2}}\left(\frac{4n}{l+1/2} -3 \right) + \frac{E_{n}^{(0)})^{2}n}{mc^{2}}\left(\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)}\right) \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left(n\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)} - \left(\frac{2n}{l+1/2} - 3/2\right)\right) \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left[n\left(\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)} - \frac{2}{l+1/2}\right) + 3/2\right] \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left[n\left(\frac{j(j+1)-l(l+1)-3/4-2l(l+1)}{l(l+1/2)(l+1)}\right) + 3/2\right] \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left[n\left(\frac{j(j+1)-3l(l+1)-3/4}{l(l+1/2)(l+1)}\right) + 3/2\right] \\
& \vdots \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left(3-\frac{4n}{j+1/2}\right) \\
& = E_{n}^{(0)}\left[1 + \frac{\alpha^{2}}{n^{2}}\left(\frac{4n}{j+1/2}-3/4\right)\right].
\end{align}
onde usamos $j=|l-s|,\cdots,l+s = l-1/2,l+1/2$ ou $l=j\mp1/2$.

### Espectros discretos degenerados
Consideremos agora o problema $H|\psi_{n}\rangle = E_{n}|\psi_{n}\rangle$ com $H=H_{0}+H_{1}$ e
\begin{equation}
H_{0}|\psi_{n,k}^{(0)}\rangle = E_{n}^{(0)}|\psi_{n,k}^{(0)}\rangle
\end{equation}
para $k=1,\cdots,d_{n}$. O desenvolvimento aqui é similar ao caso não degenerado. Consideremos um autovetor qualquer correspondente a $E_{n}^{(0)}$:
\begin{equation}
|\psi_{n}^{(0)}\rangle = \sum_{k=1}^{d_{n}}c_{n,k}|\psi_{n,k}^{(0)}\rangle.
\end{equation}
 Começamos escrevendo
\begin{align}
|\psi_{n}\rangle & = N_{n}\left(|\psi_{n}^{(0)}\rangle + |\chi_{n}\rangle \right) \\
& = N_{n}\left(\sum_{k=1}^{d_{n}}c_{n,k}|\psi_{n,k}^{(0)}\rangle + |\chi_{n}\rangle \right)
\end{align}
com $\langle\psi_{n,k}^{(0)}|\chi_{n}\rangle=0 \text{ }\forall k$. Substituindo na equação geral,
\begin{align}
& H|\psi_{n}\rangle = E_{n}|\psi_{n}\rangle \\
& (H_{0}+\xi H_{1})N_{n}\left(\sum_{k=1}^{d_{n}}c_{n,k}|\psi_{n,k}^{(0)}\rangle + |\chi_{n}\rangle \right) = E_{n}N_{n}\left(\sum_{k=1}^{d_{n}}c_{n,k}|\psi_{n,k}^{(0)}\rangle + |\chi_{n}\rangle \right) \\
& = N_{n}\left(\sum_{k=1}^{d_{n}}c_{n,k}E_{n}^{(0)}|\psi_{n,k}^{(0)}\rangle + H_{0}|\chi_{n}\rangle + \xi H_{1}\sum_{k=1}^{d_{n}}c_{n,k}|\psi_{n,k}^{(0)}\rangle + \xi H_{1}|\chi_{n}\rangle \right).
\end{align}
Dividindo esta eq. por $N_{n}$ e rearranjando,
\begin{equation}
(E_{n}-E_{n}^{(0)})\sum_{k=1}^{d_{n}}c_{n,k}|\psi_{n,k}^{(0)}\rangle = \xi \sum_{k=1}^{d_{n}}c_{n,k}H_{1}|\psi_{n,k}^{(0)}\rangle + (H_{0}+\xi H_{1}-E_{n})|\chi_{n}\rangle, \text{ (*)}
\end{equation}
e tomando o produto interno com $|\psi_{n,l}^{(0)}\rangle$ obtemos
\begin{align}
& (E_{n}-E_{n}^{(0)})\sum_{k=1}^{d_{n}}c_{n,k}\langle\psi_{n,l}^{(0)}|\psi_{n,k}^{(0)}\rangle = \xi \sum_{k=1}^{d_{n}}c_{n,k}\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle + \langle\psi_{n,l}^{(0)}|H_{0}|\chi_{n}\rangle+\xi \langle\psi_{n,l}^{(0)}|H_{1}|\chi_{n}\rangle-E_{n}\langle\psi_{n,l}^{(0)}|\chi_{n}\rangle \\
& (E_{n}-E_{n}^{(0)})\sum_{k=1}^{d_{n}}c_{n,k}\delta_{lk} = \xi \sum_{k=1}^{d_{n}}c_{n,k}\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle + E_{n}^{(0)}\langle\psi_{n,l}^{(0)}|\chi_{n}\rangle + \xi \langle\psi_{n,l}^{(0)}|H_{1}|\chi_{n}\rangle -E_{n}0 \\
& (E_{n}-E_{n}^{(0)})c_{n,l} = \xi \sum_{k=1}^{d_{n}}c_{n,k}\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle + \xi\langle\psi_{n,l}^{(0)}|H_{1}|\chi_{n}\rangle.
\end{align}

Mais uma vez, escrevemos $|\chi_{n}\rangle$ em uma base de autovetores de $H_{0}$:
\begin{align}
|\chi_{n}\rangle & = \sum_{m}\sum_{j=1}^{d_{m}}|\psi_{m,j}^{(0)}\rangle\langle\psi_{m,j}^{(0)}|\chi_{n}\rangle \\
& = \sum_{m\ne n}\sum_{j=1}^{d_{m}}\langle\psi_{m,j}^{(0)}|\chi_{n}\rangle|\psi_{m,j}^{(0)}\rangle.
\end{align}
Para obter os coeficientes tomamos o produto interno de $|\psi_{m,j}^{(0)}\rangle$ com a eq. (*) para $m\ne n$:
\begin{align}
& (E_{n}-E_{n}^{(0)})\sum_{k=1}^{d_{n}}c_{n,k}\langle\psi_{m,j}^{(0)}|\psi_{n,k}^{(0)}\rangle = \xi \sum_{k=1}^{d_{n}}c_{n,k}\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle + \langle\psi_{m,j}^{(0)}|H_{0}|\chi_{n}\rangle +\xi \langle\psi_{m,j}^{(0)}|H_{1}|\chi_{n}\rangle -E_{n}\langle\psi_{m,j}^{(0)}|\chi_{n}\rangle \\
& 0 = \xi \sum_{k=1}^{d_{n}}c_{n,k}\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle + E_{m}^{(0)}\langle\psi_{m,j}^{(0)}|\chi_{n}\rangle +\xi \langle\psi_{m,j}^{(0)}|H_{1}|\chi_{n}\rangle -E_{n}\langle\psi_{m,j}^{(0)}|\chi_{n}\rangle.
\end{align}
Para $E_{n}\ne E_{m}^{(0)}$ teremos
\begin{equation}
\langle\psi_{m,j}^{(0)}|\chi_{n}\rangle = \xi\sum_{k=1}^{d_{n}}\frac{c_{n,k}\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}} + \xi\frac{\langle\psi_{m,j}^{(0)}|H_{1}|\chi_{n}\rangle}{E_{n}-E_{m}^{(0)}}.
\end{equation}
e
\begin{equation}
|\chi_{n}\rangle = \xi\sum_{m\ne n}\sum_{j=1}^{d_{m}}\sum_{k=1}^{d_{n}}\frac{c_{n,k}\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m,j}^{(0)}\rangle + \xi\sum_{m\ne n}\sum_{j=1}^{d_{m}}\frac{\langle\psi_{m,j}^{(0)}|H_{1}|\chi_{n}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m,j}^{(0)}\rangle.
\end{equation}

Iterando,
\begin{align}
|\chi_{n}\rangle & = \xi\sum_{m\ne n}\sum_{j=1}^{d_{m}}\sum_{k=1}^{d_{n}}\frac{c_{n,k}\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m,j}^{(0)}\rangle \\
& + \xi^{2}\sum_{m\ne n}\sum_{j=1}^{d_{m}}\sum_{m'\ne n}\sum_{j'=1}^{d_{m}}\sum_{k=1}^{d_{n}}\frac{c_{n,k}\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{m',j'}^{(0)}\rangle\langle\psi_{m',j'}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle}{(E_{n}-E_{m}^{(0)})(E_{n}-E_{m'}^{(0)})}|\psi_{m,j}^{(0)}\rangle \\
& + \cdots.
\end{align}

Substituindo na eq. para a energia, usando a expansão de Raileigh-Schrödinger e a série para $1/(E_{n}-E_{m}^{(0)})$, obteremos
\begin{align}
(E_{n}-E_{n}^{(0)})c_{n,l} & = (\xi^{0}E_{n}^{(0)} + \xi^{1}E_{n}^{(1)} + \xi^{2}E_{n}^{(2)} + \xi^{3}E_{n}^{(3)} + \cdots - E_{n}^{(0)})c_{n,l} \\
& = \xi \sum_{k=1}^{d_{n}}\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle c_{n,k} \\
 & + \xi^{2}\sum_{k=1}^{d_{n}}\sum_{m\ne n}\sum_{j=1}^{d_{m}}\frac{\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{m,j}^{(0)}\rangle\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}c_{n,k} \\
 & + \xi^{3}\sum_{k=1}^{d_{n}}\sum_{m\ne n}\sum_{j=1}^{d_{m}}\sum_{m'\ne n}\sum_{j'=1}^{d_{m}}\frac{\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{m,j}^{(0)}\rangle\langle\psi_{m,j}^{(0)}|H_{1}|\psi_{m',j'}\rangle\langle\psi_{m',j'}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle}{(E_{n}-E_{m}^{(0)})(E_{n}-E_{m'}^{(0)})}c_{n,k} \\
 & + \cdots.
\end{align}

A correção perturbativa de primeira ordem,
\begin{equation}
E_{n}^{(1)}c_{n,l} = \sum_{k=1}^{d_{n}}\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle c_{n,k},
\end{equation}
pode ser escrita na forma matricial como segue
\begin{equation}
E_{n}^{(1)}\begin{bmatrix} c_{n,1} \\ c_{n,2} \\ \vdots \\ c_{n,d_{n}} \end{bmatrix} =
\begin{bmatrix} 
H_{1}^{n(1,1)} & H_{1}^{n(1,2)} & \cdots & H_{1}^{n(1,d_{n})} \\
H_{1}^{n(2,1)} & H_{1}^{n(2,2)} & \cdots & H_{1}^{n(2,d_{n})} \\
\vdots & \vdots & \ddots & \vdots \\
H_{1}^{n(d_{n},1)} & H_{1}^{n(d_{n},2)} & \cdots & H_{1}^{n(d_{n},d_{n})}
\end{bmatrix} 
\begin{bmatrix} c_{n,1} \\ c_{n,2} \\ \vdots \\ c_{n,d_{n}} \end{bmatrix},
\end{equation}
com $H_{1}^{n(j,k)}=\langle\psi_{n,j}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle$. Note que os autovalores independem da base $\{|\psi_{n,j}^{(0)}\rangle\}_{j=1}^{d_{n}}$ escolhida. Os autovetores dessa equação nos fornecem os coeficientes $c_{n,k}$ que precisaremos se quisermos calcular as correções pertubativas para os autoestados.

As correções de segunda ordem são obtidas diagonalizando (__exercício__)
\begin{equation}
\sum_{m\ne n}\sum_{j=1}^{d_{m}}\frac{H_{1}|\psi_{m,j}^{(0)}\rangle\langle\psi_{m,j}^{(0)}|H_{1}}{E_{n}^{(0)}-E_{m}^{(0)}}.
\end{equation}

#### _Exemplo_: Oscilado bidimensional carregado em um campo magnético
Consideremos um oscilador harmônico bidimensional no plano $xy$ com a partícula carregada sujeita a um campo magnético na direção $z$. O Hamiltoniano não perturbado é
\begin{align}
H_{0} & = \frac{\vec{P}^{2}}{2m} + \frac{m\omega^{2}}{2}\vec{R}^{2} \\
& = \frac{P_{1}^{2}}{2m} + \frac{m\omega^{2}}{2}X_{1}^{2} + \frac{P_{2}^{2}}{2m} + \frac{m\omega^{2}}{2}X_{2}^{2} \\
& = H_{0}^{oh1} + H_{0}^{oh2}.
\end{align}
A solução desse Hamiltoniano pode ser escrita como
\begin{equation}
H_{0}|n_{1},n_{2}\rangle = \hbar\omega(n_{1}+n_{2}+1)|n_{1},n_{2}\rangle.
\end{equation}

##### Hamiltoniano de perturbação
A partícula possui carga elétrica e gira, portanto terá um momento magnético proporcional ao momento angular:
\begin{align}
H_{1} & = -\vec{\mu}\cdot\vec{B} \\
& \propto \vec{L}\cdot\vec{B} \\
& = gL_{3},
\end{align}
em que incluímos todas as constantes de proporcionalidade em $g$. Lembrando, $L_{3}=(\vec{X}\mathrm{x}\vec{P})_{3}=\sum_{kl}\epsilon_{3kl}X_{k}P_{l}=X_{1}P_{2}-X_{2}P_{1}$. Lembrando, para o OH 1D temos $H_{0}^{ohj} = \hbar\omega(a_{j}^{\dagger}a_{j}+\mathbb{I})$ com o operador de aniquilação sendo
\begin{equation}
a_{j} = \sqrt{\frac{m\omega}{2\hbar}}\left(X_{j}+\frac{i}{m\omega}P_{j}\right).
\end{equation}
Assim
\begin{align}
X_{j} & = \sqrt{\frac{\hbar}{2m\omega}}(a_{j}^{\dagger}+a_{j}), \\
P_{j} & = i\sqrt{\frac{\hbar m\omega}{2}}(a_{j}^{\dagger}-a_{j}).
\end{align}

Escrevemos então
\begin{align}
L_{3} & = \sqrt{\frac{\hbar}{2m\omega}}(a_{1}^{\dagger}+a_{1})i\sqrt{\frac{\hbar m\omega}{2}}(a_{2}^{\dagger}-a_{2}) - i\sqrt{\frac{\hbar m\omega}{2}}(a_{2}^{\dagger}+a_{2})\sqrt{\frac{\hbar}{2m\omega}}(a_{1}^{\dagger}-a_{1}) \\
& = \frac{i\hbar}{2}\left(a_{1}^{\dagger}a_{2}^{\dagger}- a_{1}^{\dagger}a_{2} + a_{1}a_{2}^{\dagger} - a_{1}a_{2} - a_{2}^{\dagger}a_{1}^{\dagger} + a_{2}^{\dagger}a_{1} - a_{2}a_{1}^{\dagger} + a_{2}a_{1}  \right) \\
& = i\hbar(a_{2}^{\dagger}a_{1} - a_{1}^{\dagger}a_{2}).
\end{align}
Note que consideramos que os operadores de criação e de aniquilação relacionados a diferentes modos do oscilador comutam. E assim obtemos
\begin{equation}
H_{1} = gi\hbar(a_{2}^{\dagger}a_{1} - a_{1}^{\dagger}a_{2}).
\end{equation}

Vamos considerar a correção de primeira orde aplicada ao caso $n_{1}+n_{2}=2$, que corresponde ao autovalor $E_{n}^{(0)}=3\hbar\omega$ e aos autovetores $\{|0,2\rangle,|2,0\rangle,|1,1\rangle\}$. Vamos obter a matriz de $H_{1}$ nessa base do subespaço considerado. Para isso consideremos
\begin{align}
\langle n_{1},n_{2}|H_{1}|n_{1}',n_{2}'\rangle & = i\hbar g(\langle n_{1},n_{2}|a_{2}^{\dagger}a_{1}|n_{1}',n_{2}'\rangle - \langle n_{1},n_{2}|a_{1}^{\dagger}a_{2}|n_{1}',n_{2}'\rangle) \\
& = i\hbar g\left(\langle n_{1},n_{2}|a_{2}^{\dagger}\sqrt{n_{1}'}|n_{1}'-1,n_{2}'\rangle - \langle n_{1},n_{2}|a_{1}^{\dagger}\sqrt{n_{2}'}|n_{1}',n_{2}'-1\rangle\right) \\
& = i\hbar g\left(\langle n_{1},n_{2}|\sqrt{n_{2}'+1}\sqrt{n_{1}'}|n_{1}'-1,n_{2}'+1\rangle - \langle n_{1},n_{2}|\sqrt{n_{1}'+1}\sqrt{n_{2}'}|n_{1}'+1,n_{2}'-1\rangle\right) \\
& = i\hbar g\left(\sqrt{(n_{2}'+1)n_{1}'}\delta_{n_{1},n_{1}'-1}\delta_{n_{2},n_{2}'+1} - \sqrt{(n_{1}'+1)n_{2}'}\delta_{n_{1},n_{1}'+1}\delta_{n_{2},n_{2}'-1}\right).
\end{align}
Em geral, todos os termos na diagonal são nulos e aqui teremos ainda
\begin{align}
& \langle 0,2|H_{1}|2,0\rangle = 0, \\
& \langle 0,2|H_{1}|1,1\rangle = i\hbar g\left(\sqrt{(1+1)1}\delta_{0,0}\delta_{2,2} - \sqrt{(1+1)1}\delta_{0,2}\delta_{2,0}\right) = i\hbar g\sqrt{2},\\
& \langle 2,0|H_{1}|1,1\rangle = i\hbar g\left(\sqrt{(1+1)1}\delta_{2,0}\delta_{0,2} - \sqrt{(1+1)1}\delta_{2,2}\delta_{0,0}\right) = -i\hbar g\sqrt{2}.
\end{align}
Assim, na base $\{|0,2\rangle,|2,0\rangle,|1,1\rangle\}$,
\begin{equation}
H_{1} = \begin{bmatrix} 0 & 0 & i\hbar g\sqrt{2} \\ 0 & 0 & -i\hbar g\sqrt{2} \\ -i\hbar g\sqrt{2} & i\hbar g\sqrt{2}  & 0 \end{bmatrix}.
\end{equation}
De $\det(H_{1}-\lambda\mathbb{I})=0$ obtemos 
\begin{align}
& -\lambda^{3}-(i\hbar g\sqrt{2})^{2}\lambda -(-i\hbar g\sqrt{2})^{2}\lambda = 0, \\
& (\lambda^{2} - 4\hbar^{2}g^{2})\lambda = 0, \\
& \lambda = 0, \pm 2\hbar g.
\end{align}
a degenerência na energia é quebrada e teremos 
\begin{equation}
E_{2}^{(0)}\rightarrow \{\hbar(3\omega-2g),\hbar3\omega,\hbar(3\omega+2g)\}.
\end{equation}

## Teoria de perturbação dependente do tempo
Aqui queremos obter correções perturbativas para a equação de Schrödinger
\begin{equation}
i\hbar\partial_{t}|\psi_{t}\rangle = H|\psi_{t}\rangle,
\end{equation}
para $H=H_{0}+H_{1}$. 

Como um prelúdio, consideremos que $H_{0}$ é independente do tempo e que conhecemos a solução de $H_{0}|\psi_{n}^{(0)}\rangle=E_{n}^{(0)}|\psi_{t}^{(0)}\rangle$. Como nesse caso o operador de evolução temporal é $W_{t}=e^{-iH_{0}t/\hbar},$ se o estado inicial for $|\psi'\rangle=\sum_{n}c_{n}|\psi_{n}^{(0)}\rangle$ com $c_{n}=\langle\psi_{n}^{(0)}|\psi'\rangle$, o estado evoluído pode ser escrito como
\begin{align}
|\psi_{t}'\rangle & = W_{t}|\psi'\rangle \\
& = \sum_{n}c_{n}e^{-iH_{0}t/\hbar}|\psi_{n}^{(0)}\rangle \\
& = \sum_{n}c_{n}e^{-iE_{n}^{(0)}t/\hbar}|\psi_{n}^{(0)}\rangle \\
& = \sum_{n}c_{n}(t)|\psi_{n}^{(0)}\rangle.
\end{align}

Sem perda de generalidade, o estado evoluído no tempo pelo Hamiltoniano geral pode ser escrito como
\begin{align}
|\psi_{t}\rangle & = \sum_{n}A_{n}(t)|\psi_{n}^{(0)}\rangle \\
& =: \sum_{n}a_{n}(t)e^{-iE_{n}^{(0)}t/\hbar}|\psi_{n}^{(0)}\rangle.
\end{align}
Assim, se $H_{1}=\mathbb{0}$ teremos o caso particular $a_{n}(t)=c_{n}$. Nosso objetivo será obter os $a_{n}(t)$ via correções perturbativas. Usamos novamente o "truque" $H\rightarrow H_{0}+\xi H_{1}$:
\begin{align}
& i\hbar\partial_{t}\sum_{n}a_{n}(t)e^{-iE_{n}^{(0)}t/\hbar}|\psi_{n}^{(0)}\rangle = (H_{0}+\xi H_{1})\sum_{n}a_{n}(t)e^{-iE_{n}^{(0)}t/\hbar}|\psi_{n}^{(0)}\rangle, \\
& i\hbar\sum_{n}\left(\partial_{t}a_{n}(t) -i\hbar^{-1}E_{n}^{(0)}a_{n}(t)\right)e^{-iE_{n}^{(0)}t/\hbar}|\psi_{n}^{(0)}\rangle = \sum_{n}a_{n}(t)e^{-iE_{n}^{(0)}t/\hbar}(E_{n}^{(0)}\mathbb{I}+\xi H_{1})|\psi_{n}^{(0)}\rangle.
\end{align}
Seguindo, tomamos o produto interno de $|\psi_{m}^{(0)}\rangle$ com essa equação:
\begin{align}
& i\hbar\sum_{n}\left(\partial_{t}a_{n}(t) -i\hbar^{-1}E_{n}^{(0)}a_{n}(t)\right)e^{-iE_{n}^{(0)}t/\hbar}\langle \psi_{m}^{(0)}|\psi_{n}^{(0)}\rangle = \sum_{n}a_{n}(t)e^{-iE_{n}^{(0)}t/\hbar}\left( E_{n}^{(0)}\langle \psi_{m}^{(0)}|\psi_{n}^{(0)}\rangle + \xi\langle \psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle\right), \\
& i\hbar\left(\partial_{t}a_{m}(t) -i\hbar^{-1}E_{m}^{(0)}a_{m}(t)\right)e^{-iE_{m}^{(0)}t/\hbar} = a_{m}(t)e^{-iE_{m}^{(0)}t/\hbar}E_{m}^{(0)} + \xi\sum_{n}a_{n}(t)e^{-iE_{n}^{(0)}t/\hbar}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle, \\
& i\hbar\partial_{t}a_{m}(t) + E_{m}^{(0)}a_{m}(t) = a_{m}(t)E_{m}^{(0)} + \xi\sum_{n}a_{n}(t)e^{-i(E_{n}^{(0)}-E_{m}^{(0)})t/\hbar}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle, \\
& i\hbar\partial_{t}a_{m}(t) = \xi\sum_{n}a_{n}(t)e^{i\omega_{mn}t}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle.
\end{align}
onde usamos as frequências de Bohr $\omega_{m,n}=(E_{m}^{(0)}-E_{n}^{(0)})/\hbar$.

Aplicamos agora a expansão de Raileigh-Schrödinger para os coeficientes:
\begin{equation}
a_{n}(t) = \sum_{k=0}^{\infty}a_{n}^{(k)}(t)\xi^{k}.
\end{equation}
Assim segue que
\begin{align}
& i\hbar\partial_{t}\sum_{k=0}^{\infty}a_{m}^{(k)}(t)\xi^{k} = \xi\sum_{n}\sum_{k=0}^{\infty}a_{n}^{(k)}(t)\xi^{k}e^{i\omega_{mn}t}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle \\
& \sum_{k=0}^{\infty}\left(i\hbar\partial_{t}a_{m}^{(k)}(t)\right)\xi^{k} = 0\xi^{0} + \sum_{k=0}^{\infty}\left(\sum_{n}a_{n}^{(k)}(t)e^{i\omega_{mn}t}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle\right)\xi^{k+1}.
\end{align}

A correção de _ordem zero_ é obtida de
\begin{equation}
i\hbar\partial_{t}a_{m}^{(0)}(t) = 0 \Rightarrow a_{m}^{(0)}(t)=a_{m}^{(0)}(0)=c_{m}.
\end{equation}

A correção de _primeira ordem_ é obtida de
\begin{equation}
i\hbar\partial_{t}a_{m}^{(1)}(t) = \sum_{n}a_{n}^{(0)}(t)e^{i\omega_{mn}t}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle.
\end{equation}

E similarmente para um ordem qualquer $j=1,2,3,\cdots$:
\begin{equation}
i\hbar\partial_{t}a_{m}^{(j)}(t) = \sum_{n}a_{n}^{(j-1)}(t)e^{i\omega_{mn}t}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle.
\end{equation}
Aplicando o teorema fundamental do cálculo, obtemos uma equação integral que pode ser usada sucessivamente para o cálculo de correções de ordem mais alta a partir da anterior:
\begin{equation}
a_{m}^{(j)}(t) = \frac{1}{i\hbar}\int_{0}^{t}dt'\sum_{n}a_{n}^{(j-1)}(t')e^{i\omega_{mn}t'}\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle.
\end{equation}

### Exemplo: Oscilador harmônico com força externa constante
Nesse caso $H_{0}=\hbar\omega(a^{\dagger}a+\mathbb{I}/2)$, $E_{n}^{(0)}=\hbar\omega(n+1/2)$ e $|\psi_{n}^{(0)}\rangle=|n\rangle$ com $n\in\mathbb{N}$. Para ter força constante, devemos ter $U\propto X$, Nesse caso
\begin{equation}
H_{1} = -g(a^{\dagger}+a).
\end{equation}
Consideramos nesse exemplo o OH preparado inicialmente no estado fundamental $|\psi_{0}\rangle=|0\rangle=\sum_{n=0}^{\infty}c_{n}|n\rangle \Rightarrow c_{n}=\delta_{n0}$. Lembrando, $|\psi_{t}\rangle = \sum_{n=0}^{\infty}a_{n}(t)e^{i\omega_{mn}t}|n\rangle$ e $a_{n}(t)=\sum_{k=0}^{\infty}a_{n}^{(k)}$, já com o limite $\xi\rightarrow 1$ tomado.

Correção de primeira ordem:
\begin{equation}
a_{m}^{(1)}(t) = \frac{1}{i\hbar}\int_{0}^{t}dt'\sum_{n}c_{n}e^{i\omega_{mn}t'}\langle m|H_{1}|n\rangle.
\end{equation}
Agora
\begin{align}
\langle m|H_{1}|n\rangle & = -g(\langle m|a^{\dagger}|n\rangle + \langle m|a|n\rangle) \\
& = -g(\sqrt{n+1}\langle m|n+1\rangle + \sqrt{n}\langle m|n-1\rangle) \\
& = -g(\sqrt{n+1}\delta_{m,n+1} + \sqrt{n}\delta_{m,n-1}).
\end{align}
Assim
\begin{align}
a_{m}^{(1)}(t) & = \frac{-g}{i\hbar}\int_{0}^{t}dt'\sum_{n}c_{n}e^{i\omega_{mn}t'}\left(\sqrt{n+1}\delta_{m,n+1} + \sqrt{n}\delta_{m,n-1}\right) \\
& =  \frac{-g}{i\hbar}\int_{0}^{t}dt'\left(c_{m-1}e^{i\omega_{m,m-1}t'}\sqrt{m} + c_{m+1}e^{i\omega_{m,m+1}t'}\sqrt{m+1}\right) \\
& =  \frac{-g}{i\hbar}\left(\frac{\delta_{m-1,0}\sqrt{m}}{i\omega_{m,m-1}}\left[e^{i\omega_{m,m-1}t'}\right]_{0}^{t} + \frac{\delta_{m+1,0}\sqrt{m+1}}{i\omega_{m,m+1}}\left[e^{i\omega_{m,m+1}t'}\right]_{0}^{t}\right) \\
& =  \frac{-g}{i\hbar}\left(\frac{\delta_{m,1}\sqrt{1}}{i\omega_{1,1-1}}\left[e^{i\omega_{1,1-1}t}-1\right] + \frac{0\sqrt{m+1}}{i\omega_{m,m+1}}\left[e^{i\omega_{m,m+1}t}-1\right]\right) \\
& = \frac{g}{\hbar\omega_{1,0}}\left(e^{i\omega_{1,0}t}-1\right)\delta_{m,1}.
\end{align}
Se notamos que $\omega_{n,m}=(E_{n}^{(0)}-E_{m}^{(0)})/\hbar=(\hbar\omega(n+1/2)-\hbar\omega(m+1/2))/\hbar=(n-m)\omega$, a correção de primeira ordem, que é não nula somente para $m=1$, fica escrita como
\begin{equation}
a_{m}^{(1)}(t) = \frac{g}{\hbar\omega}\left(e^{i\omega t}-1\right)\delta_{m,1}.
\end{equation}

Correção de segunda ordem:
\begin{align}
a_{m}^{(2)}(t) & = \frac{1}{i\hbar}\int_{0}^{t}dt'\sum_{n}a_{n}^{(1)}(t')e^{i\omega_{mn}t'}\langle m|H_{1}|n\rangle \\
& = \frac{1}{i\hbar}\int_{0}^{t}dt'\sum_{n}\frac{F}{\hbar\omega}\left(e^{i\omega t'}-1\right)\delta_{n,1}e^{i\omega_{mn}t'}(-F)(\sqrt{n+1}\delta_{m,n+1} + \sqrt{n}\delta_{m,n-1}) \\
& = \frac{-g^{2}}{i\hbar^{2}\omega}\int_{0}^{t}dt'\left(e^{i\omega t'}-1\right)e^{i\omega_{m,1}t'}(\sqrt{1+1}\delta_{m,1+1} + \sqrt{1}\delta_{m,1-1}) \\
& = \frac{-g^{2}}{i\hbar^{2}\omega}\left(\sqrt{2}\delta_{m,2}\int_{0}^{t}dt'\left(e^{i\omega t'}-1\right)e^{i\omega_{2,1}t'} + \delta_{m,0}\int_{0}^{t}dt'\left(e^{i\omega t'}-1\right)e^{i\omega_{0,1}t'} \right) \\
& = \frac{-g^{2}}{i\hbar^{2}\omega}\left(\sqrt{2}\delta_{m,2}\int_{0}^{t}dt'\left(e^{i\omega t'}-1\right)e^{i\omega t'} + \delta_{m,0}\int_{0}^{t}dt'\left(e^{i\omega t'}-1\right)e^{-i\omega t'} \right) \\
& = \frac{-g^{2}}{i\hbar^{2}\omega}\left(\sqrt{2}\delta_{m,2}\left(\int_{0}^{t}dt'e^{2i\omega t'} - \int_{0}^{t}dt'e^{i\omega t'}\right) + \delta_{m,0}\left(\int_{0}^{t}dt' -\int_{0}^{t}dt'e^{-i\omega t'} \right)\right) \\
& = \frac{-g^{2}}{i\hbar^{2}\omega}\left(\sqrt{2}\delta_{m,2}\left(\frac{e^{2i\omega t}-1}{2i\omega} - \frac{e^{i\omega t}-1}{i\omega}\right) + \delta_{m,0}\left(t - \frac{e^{-i\omega t}-1}{-i\omega} \right)\right) \\
& = \frac{g^{2}}{\hbar^{2}\omega^{2}}\left(\sqrt{2}\delta_{m,2}2^{-1}\left(e^{2i\omega t}-1 - 2e^{i\omega t}+2\right) + \delta_{m,0}\left(i\omega t + e^{-i\omega t}-1 \right)\right) \\
& = \left(\frac{g}{\hbar\omega}\right)^{2}\left(\delta_{m,2}2^{-1/2}\left(e^{2i\omega t}-1\right)^{2} + \delta_{m,0}\left(i\omega t + e^{-i\omega t}-1 \right)\right).
\end{align}
Note que haverá correção de segunda ordem somenta para $m=0$ e para $m=2$.

Considerando forças pequenas ($g\ll \hbar\omega$), e contando que as correções de ordem maior terão magnitude proporcional a $(g/\hbar\omega)^{j}$, desprezamos correções de ordem maior que 2. Assim o estado corrigido lê-se
\begin{align}
|\psi_{t}\rangle & = \sum_{n=0}^{\infty}a_{n}(t)e^{-iE_{n}^{(0)}t/\hbar}|n\rangle \\
& \rightarrow  a_{0}(t)e^{-iE_{0}^{(0)}t/\hbar}|0\rangle + a_{1}(t)e^{-iE_{1}^{(0)}t/\hbar}|1\rangle + a_{2}(t)e^{-iE_{2}^{(0)}t/\hbar}|2\rangle \\
& \approx \left(c_{0}+a_{0}^{(1)}(t)+a_{0}^{(2)}(t)\right)e^{-i\omega t/2}|0\rangle \\ 
& + \left(c_{1}+a_{1}^{(1)}(t)+a_{1}^{(2)}(t)\right)e^{-3i\omega t/2}|1\rangle \\
& + \left(c_{2}+a_{2}^{(1)}(t)+a_{2}^{(2)}(t)\right)e^{-5i\omega t/2}|2\rangle \\
& = \left(1+0+\left(\frac{g}{\hbar\omega}\right)^{2}\left(i\omega t + e^{-i\omega t}-1 \right)\right)e^{-i\omega t/2}|0\rangle \\
& + \left(0+\frac{g}{\hbar\omega}\left(e^{i\omega t}-1\right)+0\right)e^{-3i\omega t/2}|1\rangle \\
& + \left(0+0+\left(\frac{g}{\hbar\omega}\right)^{2}2^{-1/2}\left(e^{2i\omega t}-1\right)^{2}\right)e^{-5i\omega t/2}|2\rangle \\
& = \left(1+\left(\frac{g}{\hbar\omega}\right)^{2}\left(i\omega t + e^{-i\omega t}-1 \right)\right)e^{-i\omega t/2}|0\rangle \\
& + \frac{g}{\hbar\omega}\left(e^{i\omega t}-1\right)e^{-3i\omega t/2}|1\rangle \\ 
& + \left(\frac{g}{\hbar\omega}\right)^{2}2^{-1/2}\left(e^{2i\omega t}-1\right)^{2}e^{-5i\omega t/2}|2\rangle.
\end{align}
