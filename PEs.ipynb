$1.$ As três componentes do momento angular de spin de uma partícula de spin $1$ são representadas pelos seguintes operadores Hermitianos:
\begin{equation}
S_{x} = \frac{\hbar}{\sqrt{2}}\begin{bmatrix} 0 & 1 & 0 \\ 1 & 0 & 1 \\ 0 & 1 & 0 \end{bmatrix}\text{, }
S_{y} = \frac{\hbar}{\sqrt{2}}\begin{bmatrix} 0 & -i & 0 \\ i & 0 & -i \\ 0 & i & 0 \end{bmatrix}\text{, }
S_{z} = \hbar\begin{bmatrix} 1 & 0 & 0 \\ 0 & 0 & 0 \\ 0 & 0 & -1 \end{bmatrix}. \\
\end{equation}
Essas matrizes tem espectro de autovalores $-\hbar,0,+\hbar$. Os autovetores normalizados de $S_{x}$ são
\begin{equation}
|x_{-\hbar}\rangle = \frac{1}{2}\begin{bmatrix} 1 \\ -\sqrt{2} \\ 1 \end{bmatrix}\text{, }
|x_{0}\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} -1 \\ 0 \\ 1 \end{bmatrix}\text{, }
|x_{+\hbar}\rangle = \frac{1}{2}\begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}.
\end{equation}
Dito que a componente do spin dessa partícula na direção $x$ foi medida e que foi obtido o valor $0$, calcule as probabilidades dos diferentes valores de $S_{z}$.

_Sol._ O estado preparado foi $|x_{0}\rangle$. É facil ver que os autovetores de $S_{z}$ são
\begin{equation}
|z_{-\hbar}\rangle = \begin{bmatrix} 0 \\ 0 \\ 1 \end{bmatrix}\text{, }
|z_{0}\rangle = \begin{bmatrix} 0 \\ 1 \\ 0 \end{bmatrix}\text{, }
|z_{+\hbar}\rangle = \begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}.
\end{equation}
Então, aplicando a regra de Born, teremos
\begin{align}
Pr(z_{-\hbar}|x_{0}) & = |\langle z_{-\hbar}|x_{0}\rangle|^{2} = 1/2, \\
Pr(z_{0}|x_{0}) & = |\langle z_{0}|x_{0}\rangle|^{2} = 0, \\
Pr(z_{+\hbar}|x_{0}) & = |\langle z_{+\hbar}|x_{0}\rangle|^{2} = 1/2 .
\end{align}

$2.$ Dado que os autovetores de $S_{y}$ são
\begin{equation}
|y_{-\hbar}\rangle = \frac{1}{2}\begin{bmatrix} -1 \\ i\sqrt{2} \\ 1 \end{bmatrix}\text{, }
|y_{0}\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 0 \\ 1 \end{bmatrix}\text{, }
|y_{+\hbar}\rangle = \frac{1}{2}\begin{bmatrix} -1 \\ -i\sqrt{2} \\ 1 \end{bmatrix},
\end{equation}
considere uma partícula de spin $1$ sujeita a um campo magnético constante na direção $z$ positiva e preparada inicialmente no estado $|y_{0}\rangle$. Calcule o estado evoluído no tempo, $|\psi_{t}\rangle$, e a probabilidade  $Pr(y_{0}|\psi_{t})$.

_Sol._ Colocando todas as constantes que aparecem no Hamiltoniano quantizado em $\omega$, teremos $H=-\vec{\mu}\cdot\vec{B}=-\omega S_{z}$. Como o Hamiltoniano é independente do tempo, podemos escrever
\begin{align}
U_{t} & = e^{-iHt/\hbar} = e^{\sum_{j}(-iE_{j}t/\hbar)|E_{j}\rangle\langle E_{j}|} = \sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
& = e^{i\hbar\omega t/\hbar}|z_{-\hbar}\rangle\langle z_{-\hbar}| + e^{-i0/\hbar}|z_{0}\rangle\langle z_{0}| + e^{-i\hbar\omega t/\hbar}|z_{\hbar}\rangle\langle z_{\hbar}| \\
& = e^{i\omega t}|z_{-\hbar}\rangle\langle z_{-\hbar}| + |z_{0}\rangle\langle z_{0}| + e^{-i\omega t}|z_{\hbar}\rangle\langle z_{\hbar}|.
\end{align}
Assim o estado evoluído será
\begin{align}
|\psi_{t}\rangle & = U_{t}|y_{0}\rangle \\
& = e^{i\omega t}|z_{-\hbar}\rangle\langle z_{-\hbar}|y_{0}\rangle + |z_{0}\rangle\langle z_{0}|y_{0}\rangle + e^{-i\omega t}|z_{\hbar}\rangle\langle z_{\hbar}|y_{0}\rangle \\
& = e^{i\omega t}|z_{-\hbar}\rangle(1/\sqrt{2}) + |z_{0}\rangle(0) + e^{-i\omega t}|z_{\hbar}\rangle(1/\sqrt{2}) \\
& = 2^{-1/2}\left(e^{i\omega t}|z_{-\hbar}\rangle + e^{-i\omega t}|z_{\hbar}\rangle\right).
\end{align}
Por fim
\begin{align}
Pr(y_{0}|\psi_{t}) & = |\langle y_{0}|\psi_{t}\rangle|^{2} \\
& = 2^{-1}|e^{i\omega t}\langle y_{0}|z_{-\hbar}\rangle + e^{-i\omega t}\langle y_{0}|z_{\hbar}\rangle|^{2} \\
& = 2^{-1}|e^{i\omega t}(2^{-1/2}) + e^{-i\omega t}(2^{-1/2})|^{2} \\
& = 2^{-2}|2\cos\omega t|^{2} \\
& = \cos^{2}\omega t.
\end{align}

$3.$ Considere a soma de dois momentos angulares $\vec{J}_{a}$ e $\vec{J}_{b}$ com $j_{a}=j_{b}=3/2$. Liste os possíveis valores de $m_{a}$, $m_{b}$, $j$ e $m$. Obtenha os coeficientes de Clebsch-Gordan para o estado $|j=3,m=-2\rangle$. Para um sistema quântico preparado neste estado, qual é a probabilidade de obtermos a componente $z$ de $\vec{J}_{a}$ igual a $1/2$?

_Sol_. Como $m_{s}=-j_{s},-j_{s}+1,\cdots,j_{s}$ para $s=a,b$, temos que $m_{s}=-3/2,-1/2,1/2,3/2$. Por outro lado, $j=|j_{a}-j_{b}|,|j_{a}-j_{b}|+1,\cdots,j_{a}+j_{b}$, i.e., $j=0,1,2,3$. Assim $m(j=0)=0$, $m(j=1)=-1,0,1$, $m(j=2)=-2,-1,0,1,2$ e $m(j=2)=-3,-2,-1,0,1,2,3$. Agora lembremos que $|j,m\rangle=\sum_{m_{a}=-j_{a}}^{j_{a}}\sum_{m_{b}=-j_{b}}^{j_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle$ com $m=m_{a}+m_{b}$ necessário para que os coeficientes possam ser não nulos. Vemos assim que $|j=3,m=-3\rangle=|m_{a}=-3/2\rangle\otimes|m_{b}=-3/2\rangle$. Ademais, lembrando que $J_{\pm}|j,m\rangle=\hbar\sqrt{j(j+1)-m(m\pm1)}|j,m\pm1\rangle$, notamos que
\begin{equation}
J_{+}|j=3,m=-3\rangle = \hbar\sqrt{3(3+1)-(-3)(-3+1)}|j=3,m =-3+1\rangle = \hbar\sqrt{6}|j=3,m=-2\rangle.
\end{equation}
Ou seja,
\begin{align}
|j=3,m=-2\rangle & = (\hbar\sqrt{6})^{-1} J_{+}|j=3,m=-3\rangle \\
& = (\hbar\sqrt{6})^{-1}\left(J_{+}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{+}^{b}\right)|m_{a}=-3/2\rangle\otimes|m_{b}=-3/2\rangle \\
& = (\hbar\sqrt{6})^{-1}\left(\hbar\sqrt{(3/2)(3/2+1)-(-3/2)(-3/2+1)}|m_{a}=-3/2+1\rangle\otimes|m_{b}=-3/2\rangle \\   + \hbar\sqrt{(3/2)(3/2+1)-(-3/2)(-3/2+1)}|m_{a}=-3/2\rangle\otimes|m_{b}=-3/2+1\rangle \right) \\
& = 2^{-1/2}\left(|m_{a}=-1/2\rangle\otimes|m_{b}=-3/2\rangle + |m_{a}=-3/2\rangle\otimes|m_{b}=-1/2\rangle \right).
\end{align}
A probabilidade pode ser calculada pelo módulo quadrado do coeficiente correspondente, que neste caso é $0$.

$4.$ Considere um sistema físico com Hamiltoniano $H=H_{0}+H_{1}$, onde $H_{0}=\omega S_{z}$, com $\omega\in\mathbb{R}_{>0}$, e o Hamiltoniano de perturbação é $H_{1}=\epsilon S_{x}$, com $\epsilon\in\mathbb{R}_{>0}$ e $\epsilon \ll \omega$. Dado que as expressões para as correções perturbativas de primeira e segunda ordem são obtidas, respectivamente, de $E_{n}^{(1)} = \langle E_{n}^{(0)}|H_{1}|E_{n}^{(0)}\rangle$ e $E_{n}^{(2)} = \sum_{m\ne n}|\langle E_{n}^{(0)}|H_{1}|E_{m}^{(0)}\rangle|^{2}/(E_{n}^{(0)}-E_{m}^{(0)})$, as calcule para o estado fundamental deste sistema.

_Sol._ Os autovalores do Hamiltoniano não perturbado são $E_{1}^{(0)}=-\hbar\omega$, $E_{2}^{(0)}=0$ e $E_{3}^{(0)}=\hbar\omega$. Os respectivos autovetores são $|E_{1}^{(0)}\rangle=|z_{-\hbar}\rangle$, $|E_{2}^{(0)}\rangle=|z_{0}\rangle$, $|E_{3}^{(0)}\rangle=|z_{\hbar}\rangle$. A correção de primeira ordem para o estado fundamental é
\begin{align}
E_{1}^{(1)} & = \langle E_{0}^{(0)}|H_{1}|E_{0}^{(0)}\rangle = \langle z_{-\hbar}|\epsilon S_{x}|z_{-\hbar}\rangle \propto \langle z_{-\hbar}|z_{0}\rangle = 0. 
\end{align}
Já a correção de segunda ordem para o estado fundamental é
\begin{align}
E_{0}^{(2)} & = \sum_{m\ne 1}\frac{|\langle E_{1}^{(0)}|H_{1}|E_{m}^{(0)}\rangle|^{2}}{E_{1}^{(0)}-E_{m}^{(0)}} = \frac{|\langle z_{-\hbar}|\epsilon S_{x}|z_{0}\rangle|^{2}}{-\hbar\omega-0} + \frac{|\langle z_{-\hbar}|\epsilon S_{x}|z_{\hbar}\rangle|^{2}}{-\hbar\omega-\hbar\omega} \\ 
&  = \frac{|2^{-1/2}\hbar\epsilon\langle z_{-\hbar}|(|z_{-\hbar}\rangle+|z_{\hbar}\rangle)|^{2}}{-\hbar\omega} + \frac{|2^{-1/2}\hbar\epsilon\langle z_{-\hbar}|z_{0}\rangle|^{2}}{-2\hbar\omega} \\ 
& = -\frac{\hbar\epsilon^{2}}{2\omega}.
\end{align}
