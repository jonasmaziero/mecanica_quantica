## Formalismo QuĂąntico
Em toda teoria fĂ­sica, buscamos descrever os estados possĂ­veis do sistema de interesse, as mudanĂ§as desses estados (evoluĂ§ĂŁo temporal) e as quantidades observĂĄveis e suas medidas. Por exemplo, em MecĂąnica Newtoniana (MN) o estado de um sistema em um certo instante de tempo Ă© especificado pelos valores de posiĂ§ĂŁo e velocidade em relaĂ§ĂŁo a um certo referencial inercial, e a evoluĂ§ĂŁo temporal Ă© dada pela segunda lei de Newton. Os observĂĄveis sĂŁo essencialmente as posiĂ§Ă”es, das quais obtemos as velocidades usando o parĂąmetro tempo. EstĂĄ implicito em MN que os valores dos observĂĄveis estĂŁo determinarmos antes mesmo de serem medidos, e que todos esses observĂĄveis podem ter valores bem definidos em um determinado instante de tempo. Na sequĂȘncia veremos os postulados da MecĂąnica QuĂąntica (MQ) para descrever estados, mudanĂ§as desses estados, observĂĄveis e suas medidas. Veremos que a parte da medida difere muito daquela da MN, principalmente por causa da existĂȘncia de observĂĄveis que nĂŁo podem ter valores bem definidos ao mesmo tempo.

# Postulado dos estados
O estado de um sistema quĂąntico Ă© descrito por um vetor normalizado em um espaĂ§o de Hilbert $\mathcal{H}$.

OBS. Por simplicidade, consideraremos $\mathcal{H}$ como sendo um espaĂ§o vetorial munido de uma funĂ§ĂŁo produto interno $\langle \cdot|\cdot \rangle$.

OBS. Por sistema nos referimos a um grau de liberdade. Por exemplo, o spin do elĂ©tron Ă© um sistema, outro sistema seria sua posiĂ§ĂŁo, e assim por diante. Quando consideramos dois ou mais graus de liberdade como uma Ășnica entidade, a chamamos de sistema composto.

Se dividirmos o sistema composto em subsistemas, seu estado Ă© descrito por um vetor normalizado em um espaĂ§o de Hilbert obtido tomando-se o produto tensorial dos espaĂ§os individuais $\mathcal{H} = \mathcal{H_{a}}\otimes \mathcal{H_{b}}\otimes\cdots$.

# Postulado das medidas
Quantidades observĂĄveis (mensurĂĄveis) sĂŁo representadas por operadores lineares Hermitianos definidos em $\mathcal{H}$. Ao medirmos um observĂĄvel obteremos um dos seus possĂ­veis valores, que sĂŁo representados pelos autovalores do operador Hermitiano. A MecĂąnica QuĂąntica (MQ), em geral, nĂŁo nos possibilita prever com certeza os resultados de medidas futuras. O que a MQ nos fornece sĂŁo as frequĂȘncias relativas com que esses resultados sĂŁo obtidos, que sĂŁo calculadas via a __regra de Born__:
\begin{equation}
Pr(o|\psi) = |\langle \psi |o\rangle|^{2}.
\end{equation}
Essa Ă© a probabilidade de obtermos o valor $o$ ao medirmos o observĂĄvel $O$ de um sistema que estĂĄ no estado $|\psi\rangle$ imediatamente antes da medida, e $|o\rangle$ Ă© o autovetor de $O$ correspondente ao autovalor $o$.

### Repetibilidade das medidas em MQ
Mas como sabemos em qual estado $|\psi\rangle$ o sistema foi preparado? 

Essa pergunta Ă© respondida usando a observaĂ§ĂŁo experimental de que ao medirmos um observĂĄvel duas ou mais vezes imediatamente subsequentes, obteremos o mesmo resultado. Claro, para que isso ocorra devemos ter
\begin{equation}
Pr(o|\psi)=1 \text{, o que sĂł Ă© possĂ­vel se } |\psi\rangle = |o\rangle.
\end{equation}
EntĂŁo, _postulando_ que esta varificaĂ§ĂŁo experimental parcial seja vĂĄlida de forma geral, quando dizemos que o sistema foi preparado em um certo estado, fica sub-entendido que medimos um observĂĄvel desse sistema e selecionamos o autovetor correspondente a tal estado.

### Sobre operadores Hermitianos $O = O^{\dagger}$
O autovalores desse tipo de operador sĂŁo nĂșmeros reais, e autovetores correspondentes a autovalores diferentes sĂŁo ortogonais. O conjunto de autovetores de $O$ forma uma base ortogonal para $\mathcal{H}$.

### RelaĂ§ĂŁo de completeza
Qualquer vetor do espaĂ§o vetorial $\mathcal{H}$ pode ser escrito como uma combinaĂ§ĂŁo linear dos vetores de uma base discreta e completa $\{|b_{j}\rangle\}_{j=1}^{d}$, i.e.,
\begin{equation}
|\xi\rangle = \sum_{j=1}^{d}C_{b_{j}}|b_{j}\rangle,
\end{equation}
onde $d=\dim \mathcal{H}$. Se a base $\{|b_{j}\rangle\}_{j=1}^{d}$ for ortonormal ($\langle b_{j}|b_{k}\rangle = \delta_{jk}$), pode-se verificar que os coeficientes da combinaĂ§ĂŁo linear sĂŁo dados por
\begin{equation}
C_{b_{j}}=\langle b_{j}|\xi\rangle.
\end{equation}

Nesse caso tambĂ©m podemos escrever o operador identidade em $\mathcal{H}$ como
\begin{equation}
\mathbb{I} = \sum_{j=1}^{d} |b_{j}\rangle\langle b_{j}|.
\end{equation}

Por outro lado, se temos uma base ortonormal $\{|b\rangle\}$ com incontĂĄveis vetores, a condiĂ§ĂŁo para ortonormalidade Ă© $\langle b|b'\rangle = \delta(b-b')$. O vetor $|\xi\rangle\in\mathcal{H}$ tambĂ©m pode ser escrito em termos dessa base:
\begin{equation}
|\xi\rangle = \int  db C_{b}|b\rangle.
\end{equation}
De forma anĂĄloga ao caso discreto
\begin{equation}
C_{b}=\langle b|\xi\rangle
\end{equation}
e
\begin{equation}
\int db |b\rangle\langle b| = \mathbb{I}.
\end{equation}

### RepresentaĂ§ĂŁo matricial e representaĂ§ĂŁo produto externo
Se compusermos um operador $A$ com o operador identidade, nada muda em relaĂ§ĂŁo Ă  aĂ§ĂŁo do primeiro, i.e.,
\begin{align}
A & = \mathbb{I} A \mathbb{I} = \sum_{j=1}^{d} |b_{j}\rangle\langle b_{j}| A \sum_{k=1}^{d} |b_{k}\rangle\langle b_{k}| = \sum_{j,k=1}^{d} \langle b_{j}| A |b_{k}\rangle  |b_{j}\rangle\langle b_{k}| \\
  & = \sum_{j,k=1}^{d} A_{b_{j},b_{k}}  |b_{j}\rangle\langle b_{k}|,
\end{align}
onde 
\begin{equation}
A_{b_{j},b_{k}} = \langle b_{j}| A |b_{k}\rangle \text{ Ă© a representaĂ§ĂŁo matricial do operador } A \text{ na base } \{|b_{j}\rangle\}_{j=1}^{d}.
\end{equation}
Quando um operador Ă© escrito como na equaĂ§ĂŁo acima para $A$, dizemos que o escrevemos na _forma produto externo_.

Se a base usada na relaĂ§ĂŁo de completeza for contĂ­nua,
\begin{align}
A & = \mathbb{I} A \mathbb{I} = \int db |b\rangle\langle b| A \int db' |b'\rangle\langle b'| = \iint db db' \langle b| A  |b'\rangle |b\rangle\langle b'| \\
  & = \iint db db' A_{bb'} |b\rangle\langle b'|.
\end{align}

### Valor mĂ©dio de um observĂĄvel na forma "sanduĂ­che"
Considera o valor mĂ©dio do observĂĄvel $A=\sum_{j=1}^{d} a_{j}  |a_{j}\rangle\langle a_{j}|$ para um sistema preparado no estado $|\psi\rangle$:
\begin{align}
\langle A\rangle_{\psi} & = \sum_{j=1}^{d}a_{j}Pr(a_{j}|\psi) = \sum_{j=1}^{d}a_{j}|\langle a_{j}|\psi\rangle|^{2} = \sum_{j=1}^{d}a_{j} \langle a_{j}|\psi\rangle^{*}\langle a_{j}|\psi\rangle  \\
                        & = \sum_{j=1}^{d}a_{j} \langle \psi| a_{j}\rangle\langle a_{j}|\psi\rangle = \langle \psi|\left(\sum_{j=1}^{d}a_{j} | a_{j}\rangle\langle a_{j}|\right)|\psi\rangle  \\
                        & = \langle \psi|A|\psi\rangle.
\end{align}

### Amplitudes de probabilidade e normalizaĂ§ĂŁo
Os autovetores de um observĂĄvel qualquer formam uma base ortogonal para pelo menos um sub-espaĂ§o do espaĂ§o de estados do sistema. Vamos assumir aqui que essa base Ă© completa para todo esse espaĂ§o. Ă frequentemente Ăștil escrevermos o estado do sistema imediatamente antes da medida, $|\psi\rangle$, como uma combinaĂ§ĂŁo linear dos autovetores do observĂĄvel $O$ que vai ser medido:
\begin{equation}
|\psi\rangle = \sum_{j=1}^{d} C_{o_{j}} |o_{j}\rangle \text{ no caso discreto} \text{ ou } |\psi\rangle = \int do C_{o} |o\rangle  \text{ no caso contĂ­nuo}.
\end{equation}

Se colocarmos estes vetores de estado na regra de Born, veremos que a probabilidade de obsermos o valor $o_{j}$ Ă©
\begin{equation}
Pr(o_{j}|\psi) = |\langle o_{j}|\psi\rangle|^{2} = \left|\langle o_{j}|\sum_{k=1}^{d} C_{o_{k}} |o_{k}\rangle\right|^{2} = \left|\sum_{k=1}^{d} C_{o_{k}} \langle o_{j}|o_{k}\rangle\right|^{2} = \left|\sum_{k=1}^{d} C_{o_{k}} \delta_{jk} \right|^{2} = \left|C_{o_{j}}\right|^{2}.
\end{equation}
Para espectros contĂ­nuos, obteremos a densidade de probabilidade
\begin{equation}
\rho(o|\psi) = \left|\langle b|\psi\rangle\right|^{2} = \left|\langle b|\int db' C_{b'} |b'\rangle\right|^{2} = \left|\int db' C_{b'} \langle b|b'\rangle\right|^{2} = \left|\int db' C_{b'} \delta(b-b')\right|^{2} = \left|C_{b}\right|^{2}.
\end{equation}
Com isso fica claro o motivo para chamarmos $C_{o_{j}}$ e $C_{o}$ de amplitudes de probabilidade, o mĂłdulo quadrado dessas quantidades Ă© uma probabilidade e uma densidade de probabilidade, respectivamente. 

Aqui tambĂ©m podemos ver o motivo para exigirmos que os vetores de estado sejam normalizados:
\begin{align}
||\psi||^{2} & = \langle\psi|\psi\rangle = \sum_{j=1}^{d} C_{o_{j}}^{*} \langle o_{j}| \sum_{k=1}^{d} C_{o_{k}} |o_{k}\rangle = \sum_{j,k=1}^{d} C_{o_{j}}^{*}C_{o_{k}} \langle o_{j}|o_{k}\rangle = \sum_{j,k=1}^{d} C_{o_{j}}^{*}C_{o_{k}} \delta_{jk} \\ 
& = \sum_{j=1}^{d} |C_{o_{j}}|^2 = \sum_{j=1}^{d} Pr(o_{j}|\psi) := 1.
\end{align}
e
\begin{align}
||\psi||^{2} & = \langle\psi|\psi\rangle  = \int do C_{o}^{*} \langle o| \int do' C_{o'} |o'\rangle = \iint do do' C_{o}^{*}C_{o'} \langle o|o'\rangle = \iint do do' C_{o}^{*}C_{o'} \delta(o-o') \\
             & = \int do |C_{o}|^{2} = \int do \rho(o|\psi) := 1.
\end{align}

### Pulos quĂąnticos
Como dito acima, quando medimos um observĂĄvel $O$ e obtemos um dos seus possĂ­veis valores, preparamos o sistema no vetor de estado igual ao autovetor correspondente ao valor obtido na medida. Ilustraremos esse efeito pelas equaĂ§Ă”es:
\begin{eqnarray}
|\psi\rangle & = & \sum_{j=1}^{d} C_{o_{j}} |o_{j}\rangle \xrightarrow[]{\text{obtĂ©m } o_{k}} |o_{k}\rangle, \\
|\psi\rangle & = & \int do C_{o} |o\rangle  \xrightarrow[]{\text{obtĂ©m } o'} |o'\rangle.
\end{eqnarray}
Em fĂ­sica clĂĄssica, estamos acostumados com mudanĂ§as contĂ­nuas de estado (de posiĂ§Ă”es e velocidades, por exemplo). Na MQ pode ocorrer do sistema pular "instantaneamente" de uma configuraĂ§ĂŁo para outra que estĂĄ a uma "distĂąncia finita" da primeira. Ou seja, nĂŁo podemos traĂ§ar uma trajetĂłria no espaĂ§o de estados quĂąnticos para descrever o que ocorre durante o processo de medida.

## Incompatibilidade quĂąntica
### __Teorema__: 
Dois operadores Hermitianos $A$ e $B$ compartilham o mesmo conjunto de autovetores se e somente se comutarem.
### Prova:
Seja $A = \sum_{j=1}^{d}a_{j}|\alpha_{j}\rangle\langle \alpha_{j}|$ e $B = \sum_{j=1}^{d}b_{j}|\beta_{j}\rangle\langle \beta_{j}|$, com $d:=\dim\mathcal{H}$. Note que por causa da degenerescĂȘncia, podemos ter e.g. $a_{1}=a_{2}$ ou e.g. $b_{2}=b_{3}=b_{4}$. ComeĂ§amos assumindo que os dois operadores compartilham a mesma base de autovetores, i.e., $|\alpha_{j}\rangle\equiv|\beta_{j}\rangle:=|j\rangle$. Nesse caso podemos escrever as representaĂ§Ă”es produto externo: $A = \sum_{j=1}^{d} a_{j}|j\rangle\langle j|$ e $B = \sum_{j=1}^{d} b_{j}|j\rangle\langle j|.$ O comutador Ă© entĂŁo
\begin{align}
[A,B] & = AB - BA = \sum_{j=1}^{d} a_{j}|j\rangle\langle j|\sum_{k=1}^{d} b_{k}|k\rangle\langle k| - \sum_{k=1}^{d} b_{k}|k\rangle\langle k|\sum_{j=1}^{d} a_{j}|j\rangle\langle j| \\
      & = \sum_{j,k=1}^{d} a_{j}b_{k}|j\rangle\langle j|k\rangle\langle k| - \sum_{k,j=1}^{d} b_{k}a_{j}|k\rangle\langle k|j\rangle\langle j| = \sum_{j,k=1}^{d} a_{j}b_{k}|j\rangle\langle k|\delta_{jk} - \sum_{k,j=1}^{d} b_{k}a_{j}|k\rangle\langle j|\delta_{jk} \\
      & = \sum_{j=1}^{d} a_{j}b_{j}|j\rangle\langle j| - \sum_{j=1}^{d} b_{j}a_{j}|j\rangle\langle j| = 0.
\end{align}

Seguindo com a segunda parte da prova, assumimos que os operadores comutam: $[A,B] = 0 \Rightarrow AB-BA=0 \Rightarrow AB=BA$. Vem assim que $\forall j$:
\begin{align}
A(B|\alpha_{j}\rangle) & = (AB)|\alpha_{j}\rangle = (BA)|\alpha_{j}\rangle = B(A|\alpha_{j}\rangle) = B (a_{j}|\alpha_{j}\rangle) \\
              & = a_{j}(B|\alpha_{j}\rangle). 
\end{align}
Ou seja, nesse caso devemos ter que $B|\alpha_{j}\rangle$ e $|\alpha_{j}\rangle$ sĂŁo ambos autovetores de $A$ correspondentes ao mesmo autovalor $a_{j}$. Se $a_{j}$ Ă© degenerado, existe um subespaĂ§o de $\mathcal{H}$ gerado por autovetores de $A$ correspondentes ao autovalor $a_{j}$. Chamemos esse subespaĂ§o de $\mathcal{H}_{j}$. Seja $\{|\alpha_{k}^{j}\rangle\}_{k=1}^{d_{j}}$ uma base de $\mathcal{H}_{j}$, com $d_{j}=\dim(\mathcal{H}_{j})$. Temos entĂŁo que $A|\alpha_{k}^{j}\rangle=a_{j}|\alpha_{k}^{j}\rangle$ para $k=1,\cdots,d_{j}$. AlĂ©m disso, $|\alpha_{j}\rangle\in\mathcal{H}_{j}$. Como $B|\alpha_{j}\rangle\in\mathcal{H}_{j}$, vemos que $B:\mathcal{H}_{j}\rightarrow \mathcal{H}_{j}$. Como $B$ Ă© um operador Hermitiano, existe uma base de $\mathcal{H}_{j}$ que o diagonaliza: $B|\alpha_{k}^{j}\rangle=b_{j}|\alpha_{k}^{j}\rangle$, com $j=1,\cdots,d_{j}$. Demonstramos com isso que $[A,B] = 0$ implica que $A$ e $B$ compartilham a mesma base de autovetores, concluindo assim a prova do teorema nesse caso.

### Incompatibilidade e superposiĂ§ĂŁo quĂąntica

Com a prova deste teorema sabemos que se dois observĂĄveis $A$ e $B$ de um sistema nĂŁo comutam, eles nĂŁo podem compartilhar todos os autovetores. Se, medindo o observĂĄvel $A$, preparamos o sistema em um estado que nĂŁo Ă© autovetor do observĂĄvel $B$, este vetor deve ser obrigatoriamente uma combinaĂ§ĂŁo linear dos autovetores de $B$ com duas ou mais amplitudes de probabilidade nĂŁo nulas. Por exemplo, se $B|a_{3}\rangle$ nĂŁo Ă© proporcional a $|a_{3}\rangle$ teremos e.g.
\begin{equation}
|?\rangle \xrightarrow[\text{sai } a_{3}]{\text{mede } A} |a_{3}\rangle = \sqrt{1/4}|b_{2}\rangle +  \sqrt{3/4}|b_{5}\rangle  \xrightarrow[]{\text{mede } B} \text{sai }  b_{2} \text{ com prob. 1/4 ou } b_{5} \text{ com prob. 3/4}.
\end{equation}

# Postulado da dinĂąmica
As mudanĂ§as de estados em MQ sĂŁo descritas por uma transformaĂ§ĂŁo unitĂĄria:
\begin{equation}
|\psi_{t}\rangle = U_{t}|\psi_{0}\rangle,
\end{equation}
i.e., $U_{t}U_{t}^{\dagger}=U_{t}U_{t}^{\dagger}=\mathbb{I}$. O operador de evoluĂ§ĂŁo temporal sendo obtido da __equaĂ§ĂŁo de SchrĂ¶dinger__:
\begin{equation}
i\hbar\partial_{t} U_{t} = H_{t}U_{t},
\end{equation}
onde $H_{t}$ Ă© o operador Hamiltoniano do sistema no instante de tempo $t$, $\hbar$ Ă© a constante de Planck dividida por $2\pi$ e usamos a notaĂ§ĂŁo $\partial_{t} = \frac{\partial}{\partial_{t}}$.

_OBS._ As duas equaĂ§Ă”es acima sĂŁo equivalentes Ă  equaĂ§ĂŁo de SchrĂ¶dinger para o estado
\begin{equation}
i\hbar\partial_{t} |\psi_{t}\rangle = H_{t}|\psi_{t}\rangle.
\end{equation}
Para verificar isso nota que
\begin{align}
i\hbar\partial_{t} |\psi_{t}\rangle & = i\hbar\partial_{t}(U_{t}|\psi_{0}\rangle) = (i\hbar\partial_{t}U_{t})|\psi_{0}\rangle = H_{t}U_{t}|\psi_{0}\rangle \\
                                    & = H_{t}|\psi_{t}\rangle.
\end{align}

### SoluĂ§ĂŁo para $U_{t}$ quando $H_{t}=H$
Nesse tipo de situaĂ§ĂŁo, i.e.,quando o Hamiltoniano nĂŁo depende do tempo, podemos verificar que a seguinte forma geral Ă© vĂĄlida:
\begin{equation}
U_{t} = e^{-iHt/\hbar}.
\end{equation}
Faremos isso primeiramente usando a decomposiĂ§ĂŁo espectral do Hamiltoniano, $H=\sum_{j}E_{j}|E_{j}\rangle\langle E_{j}|$, e aplicando a funĂ§ĂŁo de operador:
\begin{align}
i\hbar\partial_{t} U_{t} & = i\hbar\partial_{t}e^{-iHt/\hbar} = i\hbar\partial_{t}\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| = i\hbar\sum_{j}\partial_{t}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
& = i\hbar\sum_{j}(-i/\hbar)E_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}|  = \sum_{j}e^{-iE_{j}t/\hbar}E_{j}|E_{j}\rangle\langle E_{j}|  \\
& = \sum_{j}e^{-iE_{j}t/\hbar}H|E_{j}\rangle\langle E_{j}| = H\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}|  = H U_{t}.
\end{align}

### SoluĂ§ĂŁo para $U_{t}$ quando $[H_{t},H_{t'}]=\mathbb{0}$
Vamos verificar que neste caso
\begin{equation}
U_{t} = e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}.
\end{equation}
Teremos
\begin{align}
i\hbar\partial_{t} U_{t} & = i\hbar\partial_{t} e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}  = i\hbar\partial_{t}\sum_{j=0}^{\infty}\frac{1}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j} \\
                         & = i\hbar\partial_{t}\left[\mathbb{I} + \sum_{j=1}^{\infty}\frac{1}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j}\right] \\
                         & = i\hbar\left[\mathbb{0} + \sum_{j=1}^{\infty}\frac{j}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j-1}\partial_{t}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)\right] \\
                         & = i\hbar\frac{-i}{\hbar}\partial_{t}\left(\int_{0}^{t}H_{t'}dt'\right)\sum_{j=1}^{\infty}\frac{j}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j-1},
\end{align}
No Ășltimo passo usamos a comutatividade do Hamiltoniano em tempos diferentes. Seguindo, aplicando o teorema fundamental do cĂĄlculo e fazendo a mudanĂ§a de variĂĄvel $k=j-1$ obtemos
\begin{align}
i\hbar\partial_{t} U_{t} & = H_{t}\sum_{k=0}^{\infty}\frac{(k+1)}{(k+1)!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{k} = H_{t}\sum_{k=0}^{\infty}\frac{1}{k!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{k} \\ 
                         & = H_{t}e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'} = H_{t}U_{t}.
\end{align}
