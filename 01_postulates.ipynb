# Formalismo Quântico
Em toda teoria física, buscamos descrever os estados possíveis do sistema de interesse, as mudanças desses estados (evolução temporal) e as quantidades observáveis e suas medidas. Por exemplo, em Mecânica Newtoniana (MN) o estado de um sistema é especificado pelos valores de posição e velocidade em relação a um certo referencial inercial, e a evolução temporal é dada pela segunda lei de Newton. Os observáveis são essencialmente as posições, das quais obtemos as velocidades usando o parâmetro tempo. Está implicito em MN que os valores dos observáveis estão determinarmos antes mesmo de serem medidos, e que todos esses observáveis podem ter valores bem definidos em um determinado instante de tempo. Na sequência veremos os postulados da Mecânica Quântica (MQ) para descrever estados, mudanças desses estados, observáveis e suas medidas. Veremos que a parte da medida difere muito daquela da MN, principalmente por causa da existência de observáveis que não podem ter valores bem definidos ao mesmo tempo.

## Postulado dos estados
O estado de um sistema quântico é descrito por um vetor normalizado em um espaço de Hilbert $\mathcal{H}$.

OBS. Por simplicidade, consideraremos $\mathcal{H}$ como sendo um espaço vetorial munido de uma função produto interno $\langle \cdot|\cdot \rangle$.

OBS. Por sistema nos referimos a um grau de liberdade. Por exemplo, o spin do elétron é um sistema, outro sistema seria sua posição, e assim por diante. Quando consideramos dois ou mais graus de liberdade como uma única entidade, a chamamos de sistema composto.

Se dividirmos o sistema composto em subsistemas, seu estado é descrito por um vetor normalizado em um espaço de Hilbert obtido tomando-se o produto tensorial dos espaços individuais $\mathcal{H} = \mathcal{H_{a}}\otimes \mathcal{H_{b}}\otimes\cdots$.

## Postulado das medidas
Quantidades observáveis (mensuráveis) são representadas por operadores lineares Hermitianos definidos em $\mathcal{H}$. Ao medirmos um observável obteremos um dos seus possíveis valores. A Mecânica Quântica (MQ), em geral, não nos possibilita prever com certeza os resultados de medidas futuras. O que a MQ nos fornece são as frequências relativas com que esses resultados são obtidos, que são calculadas via a __regra de Born__:
\begin{equation}
Pr(o|\psi) = |\langle \psi |o\rangle|^{2}.
\end{equation}
Essa é a probabilidade de obtermos o valor $o$ ao medirmos o observável $O$ de um sistema que está no estado $|\psi\rangle$ imediatamente antes da medida, e $|o\rangle$ é o autovetor de $O$ correspondente ao autovalor $o$.

### Repetibilidade das medidas em MQ
Mas como sabemos em qual estado $|\psi\rangle$ o sistema foi preparado? 

Essa pergunta é respondida usando a observação experimental de que ao medirmos um observável duas ou mais vezes imediatamente subsequentes, obteremos o mesmo resultado. Claro, para que isso ocorra devemos ter $Pr(o|\psi)=1$, o que só é possível se $|\psi\rangle = |o\rangle$. Então, _postulando_ que esta varificação experimental parcial seja válida de forma geral, quando dizemos que o sistema foi preparado em um certo estado, fica sub-entendido que medimos um observável desse sistema e selecionamos o autovetor correspondente a tal estado.

### Sobre operadores Hermitianos $O = O^{\dagger}$
O autovalores desse tipo de operador são números reais, e autovetores correspondentes a autovalores diferentes são ortogonais. O conjunto de autovetores de $O$ forma uma base ortogonal para $\mathcal{H}$.

### Relação de completeza
Qualquer vetor do espaço vetorial $\mathcal{H}$ pode ser escrito como uma combinação linear dos vetores de uma base discreta e completa $\{|b_{j}\rangle\}_{j=1}^{d}$, i.e.,
\begin{equation}
|\xi\rangle = \sum_{j=1}^{d}C_{b_{j}}|b_{j}\rangle,
\end{equation}
onde $d=\dim \mathcal{H}$. Se a base $\{|b_{j}\rangle\}_{j=1}^{d}$ for ortonormal ($\langle b_{j}|b_{k}\rangle = \delta_{jk}$), pode-se verificar que os coeficientes da combinação linear são dados por (__exercício__)
\begin{equation}
C_{b_{j}}=\langle b_{j}|\xi\rangle.
\end{equation}

Nesse caso também podemos escrever o operador identidade em $\mathcal{H}$ como (__exercício__)
\begin{equation}
\mathbb{I} = \sum_{j=1}^{d} |b_{j}\rangle\langle b_{j}|.
\end{equation}

Por outro lado, se temos uma base ortonormal $\{|b\rangle\}$ com incontáveis vetores, a condição para ortonormalidade é $\langle b|b'\rangle = \delta(b-b')$. O vetor $|\xi\rangle\in\mathcal{H}$ também pode ser escrito em termos dessa base:
\begin{equation}
|\xi\rangle = \int  db C_{b}|b\rangle.
\end{equation}
De forma análoga ao caso discreto (__exercício__)
\begin{equation}
C_{b}=\langle b|\xi\rangle
\end{equation}
e (__exercício__)
\begin{equation}
\int db |b\rangle\langle b| = \mathbb{I}.
\end{equation}

### Representação matricial e representação produto externo
Se compusermos um operador $A$ com o operador identidade, nada muda em relação à ação do primeiro, i.e.,
\begin{align}
A & = \mathbb{I} A \mathbb{I} \\
  & = \sum_{j=1}^{d} |b_{j}\rangle\langle b_{j}| A \sum_{k=1}^{d} |b_{k}\rangle\langle b_{k}| \\
  & = \sum_{j,k=1}^{d} \langle b_{j}| A |b_{k}\rangle  |b_{j}\rangle\langle b_{k}| \\
  & = \sum_{j,k=1}^{d} A_{b_{j},b_{k}}  |b_{j}\rangle\langle b_{k}|,
\end{align}
onde $A_{b_{j},b_{k}} = \langle b_{j}| A |b_{k}\rangle$ é a _representação matricial_ do operador $A$ na base $\{|b_{j}\rangle\}_{j=1}^{d}$. Quando um operador é escrito como na última equação para $A$, dizemos que o escrevemos na _forma produto externo_.

Se a base usada na relação de completeza for contínua,
\begin{align}
A & = \mathbb{I} A \mathbb{I} \\
  & = \int db |b\rangle\langle b| A \int db' |b'\rangle\langle b'| \\
  & = \iint db db' \langle b| A  |b'\rangle |b\rangle\langle b'| \\
  & = \iint db db' A_{bb'} |b\rangle\langle b'|.
\end{align}

__Exercício:__
Mostre que se usarmos a base de autovetores de $A$ para representá-lo, teremos que $\langle a_{j} |A|a_{k}\rangle = a_{j}\delta_{jk}$ e $\langle a|A|a'\rangle = a\delta(a-a')$ e portanto que $A=\sum_{j=1}^{d} a_{j}  |a_{j}\rangle\langle a_{j}|$ e $A=\int da a |a\rangle\langle a|$.

### Valor médio de um observável na forma "sanduíche"
Considera o valor médio do observável $A=\sum_{j=1}^{d} a_{j}  |a_{j}\rangle\langle a_{j}|$ para um sistema preparado no estado $|\psi\rangle$:
\begin{align}
\langle A\rangle_{\psi} & = \sum_{j=1}^{d}a_{j}Pr(a_{j}|\psi) \\
                        & = \sum_{j=1}^{d}a_{j}|\langle a_{j}|\psi\rangle|^{2} \\
                        & = \sum_{j=1}^{d}a_{j} \langle a_{j}|\psi\rangle^{*}\langle a_{j}|\psi\rangle  \\
                        & = \sum_{j=1}^{d}a_{j} \langle \psi| a_{j}\rangle\langle a_{j}|\psi\rangle  \\
                        & = \langle \psi|\left(\sum_{j=1}^{d}a_{j} | a_{j}\rangle\langle a_{j}|\right)|\psi\rangle  \\
                        & = \langle \psi|A|\psi\rangle.
\end{align}

__Exercício:__ Mostre que a mesma relação é obtida se $A$ tiver espectro contínuo.

### Amplitudes de probabilidade e normalização
Os autovetores de um observável qualquer formam uma base ortogonal para pelo menos um sub-espaço do espaço de estados do sistema. Vamos assumir aqui que essa base é completa para todo esse espaço. É frequentemente útil escrevermos o estado do sistema imediatamente antes da medida, $|\psi\rangle$, como uma combinação linear dos autovetores do observável $O$ que vai ser medido:
\begin{equation}
|\psi\rangle = \sum_{j=1}^{d} C_{o_{j}} |o_{j}\rangle \text{ no caso discreto}
\end{equation}
ou
\begin{equation}
|\psi\rangle = \int do C_{o} |o\rangle  \text{ no caso contínuo}.
\end{equation}

Se colocarmos estes vetores de estado na regra de Born, veremos que a probabilidade de obsermos o valor $o_{j}$ é
\begin{align}
Pr(o_{j}|\psi) & = |\langle o_{j}|\psi\rangle|^{2} \\ 
               & = \left|\langle o_{j}|\sum_{k=1}^{d} C_{o_{k}} |o_{k}\rangle\right|^{2} \\
               & = \left|\sum_{k=1}^{d} C_{o_{k}} \langle o_{j}|o_{k}\rangle\right|^{2} \\
               & = \left|\sum_{k=1}^{d} C_{o_{k}} \delta_{jk} \right|^{2} \\
               & = \left|C_{o_{j}}\right|^{2}.
\end{align}
Para espectros contínuos, obteremos a densidade de probabilidade
\begin{align}
\rho(o|\psi) & = \left|\langle b|\psi\rangle\right|^{2} \\
             & = \left|\langle b|\int db' C_{b'} |b'\rangle\right|^{2} \\
             & = \left|\int db' C_{b'} \langle b|b'\rangle\right|^{2} \\
             & = \left|\int db' C_{b'} \delta(b-b')\right|^{2} \\
             & = \left|C_{b}\right|^{2}.
\end{align}
Com isso fica claro o motivo para chamarmos $C_{o_{j}}$ e $C_{o}$ de amplitudes de probabilidade. 

Aqui também podemos ver o motivo para exigirmos que os vetores de estado sejam normalizados:
\begin{align}
||\psi||^{2} & = \langle\psi|\psi\rangle \\ 
             & = \sum_{j=1}^{d} C_{o_{j}}^{*} \langle o_{j}| \sum_{k=1}^{d} C_{o_{k}} |o_{k}\rangle \\
             & = \sum_{j,k=1}^{d} C_{o_{j}}^{*}C_{o_{k}} \langle o_{j}|o_{k}\rangle \\
             & = \sum_{j,k=1}^{d} C_{o_{j}}^{*}C_{o_{k}} \delta_{jk} \\
             & = \sum_{j=1}^{d} |C_{o_{j}}|^2 \\
             & = \sum_{j=1}^{d} Pr(o_{j}|\psi) \\
             & = 1.
\end{align}
e
\begin{align}
||\psi||^{2} & = \langle\psi|\psi\rangle \\ 
             & = \int do C_{o}^{*} \langle o| \int do' C_{o'} |o'\rangle \\
             & = \iint do do' C_{o}^{*}C_{o'} \langle o|o'\rangle \\
             & = \iint do do' C_{o}^{*}C_{o'} \delta(o-o') \\
             & = \int do |C_{o}|^{2} \\
             & = \int do \rho(o|\psi) \\
             & = 1.
\end{align}

### Pulos quânticos
Como dito acima, quando medimos um observável $O$ e obtemos um dos seus possíveis valores, preparamos o sistema no vetor de estado igual ao autovetor correspondente ao valor obtido na medida. Ilustraremos esse efeito pelas equações:
\begin{eqnarray}
|\psi\rangle & = & \sum_{j=1}^{d} C_{o_{j}} |o_{j}\rangle \xrightarrow[]{\text{obtém } o_{k}} |o_{k}\rangle, \\
|\psi\rangle & = & \int do C_{o} |o\rangle  \xrightarrow[]{\text{obtém } o'} |o'\rangle.
\end{eqnarray}
Em física clássica, estamos acostumados com mudanças contínuas de estado (de posições e velocidades, por exemplo). Na MQ pode ocorrer do sistema pular "instantaneamente" de uma configuração para outra que está a uma "distância finita" da primeira. Ou seja, não podemos traçar uma trajetória no espaço de estados quânticos para descrever o que ocorre durante o processo de medida.

### Incompatibilidade quântica
#### __Teorema__: 
Dois operadores Hermitianos $A$ e $B$ compartilham o mesmo conjunto de autovetores se e somente se comutarem.
#### Prova:

Começaremos assumindo _espectro discreto não degenerado_ e que os dois operadores compartilham a mesma base de autovetores. Nesse caso podemos escrever as representações produto externo: $A = \sum_{j=1}^{d} a_{j}|j\rangle\langle j|$ e $B = \sum_{j=1}^{d} b_{j}|j\rangle\langle j|.$ O comutador é então
\begin{align}
[A,B] & = AB - BA \\
      & = \sum_{j=1}^{d} a_{j}|j\rangle\langle j|\sum_{k=1}^{d} b_{k}|k\rangle\langle k| - \sum_{k=1}^{d} b_{k}|k\rangle\langle k|\sum_{j=1}^{d} a_{j}|j\rangle\langle j| \\
      & = \sum_{j,k=1}^{d} a_{j}b_{k}|j\rangle\langle j|k\rangle\langle k| - \sum_{k,j=1}^{d} b_{k}a_{j}|k\rangle\langle k|j\rangle\langle j| \\
      & = \sum_{j,k=1}^{d} a_{j}b_{k}|j\rangle\langle k|\delta_{jk} - \sum_{k,j=1}^{d} b_{k}a_{j}|k\rangle\langle j|\delta_{jk} \\
      & = \sum_{j=1}^{d} a_{j}b_{j}|j\rangle\langle j| - \sum_{j=1}^{d} b_{j}a_{j}|j\rangle\langle j| \\
      & = 0.
\end{align}
Seguindo com a segunda parte da prova, assumimos que os operadores comutam: $[A,B] = 0 \Rightarrow AB-BA=0 \Rightarrow AB=BA$. Consideremos que $\{|j\rangle\}_{j=1}^{d}$ é a base de autovetores de $A$. Vem assim que $\forall j$
\begin{align}
A(B|j\rangle) & = (AB)|j\rangle = (BA)|j\rangle \\
              & = B(A|j\rangle) = B (a_{j}|j\rangle) \\
              & = a_{j}(B|j\rangle). 
\end{align}
Ou seja, nesse caso devemos ter que $B|j\rangle$ e $|j\rangle$ são ambos autovetores de $A$ correspondentes ao mesmo autovalor. Mas como $A$ é não degenerado devemos ter que $B|j\rangle$ e $|j\rangle$ estão na mesma "direção" (são proporcionais). Mas isso implica que $|j\rangle$ também é autovetor de $B$:
\begin{equation}
B|j\rangle \propto |j\rangle \Rightarrow B|j\rangle = b_{j} |j\rangle.
\end{equation}
Demonstramos com isso que $[A,B] = 0$ implica que $A$ e $B$ compartilham a mesma base de autovetores, concluindo assim a prova do teorema nesse caso.

O análogo segue para $B$.

### Incompatibilidade e superposição quântica

Com a prova deste teorema sabemos que se dois observáveis $A$ e $B$ de um sistema não comutam, eles não podem compartilhar todos os autovetores. Se, medindo o observável $A$, preparamos o sistema em um estado que não é autovetor do observável $B$, este vetor deve ser obrigatoriamente uma combinação linear dos autovetores de $B$ com duas ou mais amplitudes de probabilidade não nulas. Por exemplo, se $B|a_{3}\rangle$ não é proporcional a $|a_{3}\rangle$ teremos e.g.
\begin{equation}
|?\rangle \xrightarrow[\text{sai } a_{3}]{\text{mede } A} |a_{3}\rangle = \sqrt{1/4}|b_{2}\rangle +  \sqrt{3/4}|b_{5}\rangle  \xrightarrow[]{\text{mede } B} \text{sai }  b_{2} \text{ com prob. 1/4 ou } b_{5} \text{ com prob. 3/4}.
\end{equation}

## Postulado da dinâmica
As mudanças de estados em MQ são dada por uma transformação unitária
\begin{equation}
|\psi_{t}\rangle = U_{t}|\psi_{0}\rangle,
\end{equation}
com o operador de evolução temporal sendo obtido da __equação de Schrödinger__
\begin{equation}
i\hbar\partial_{t} U_{t} = H_{t}U_{t},
\end{equation}
onde $H_{t}$ é o operador Hamiltoniano do sistema no instante de tempo $t$, $\hbar$ é a constante de Planck dividida por $2\pi$ e usamos a notação $\partial_{t} = \frac{\partial}{\partial_{t}}$.

_OBS._ As duas equações acima são equivalentes à equação de Schrödinger para o estado
\begin{equation}
i\hbar\partial_{t} |\psi_{t}\rangle = H_{t}|\psi_{t}\rangle.
\end{equation}
Para verificar isso nota que
\begin{align}
i\hbar\partial_{t} |\psi_{t}\rangle & = i\hbar\partial_{t}(U_{t}|\psi_{0}\rangle) \\ 
                                    & = (i\hbar\partial_{t}U_{t})|\psi_{0}\rangle \\
                                    & = H_{t}U_{t}|\psi_{0}\rangle \\
                                    & = H_{t}|\psi_{t}\rangle.
\end{align}

### Solução para $U_{t}$ quando $H_{t}=H$
Nesse tipo de situação, i.e.,quando o Hamiltoniano não depende do tempo, podemos verificar que a seguinte forma geral é válida:
\begin{equation}
U_{t} = e^{-iHt/\hbar}.
\end{equation}
Faremos isso primeiramente usando a decomposição espectral do Hamiltoniano, $H=\sum_{j}E_{j}|E_{j}\rangle\langle E_{j}|$, e aplicando a função de operador:
\begin{align}
i\hbar\partial_{t} U_{t} & = i\hbar\partial_{t}e^{-iHt/\hbar} \\
                         & = i\hbar\partial_{t}\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = i\hbar\sum_{j}\partial_{t}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = i\hbar\sum_{j}(-i/\hbar)E_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = \sum_{j}e^{-iE_{j}t/\hbar}E_{j}|E_{j}\rangle\langle E_{j}| \\
                         & = \sum_{j}e^{-iE_{j}t/\hbar}H|E_{j}\rangle\langle E_{j}| \\
                         & = H\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = H U_{t}.
\end{align}

### Solução para $U_{t}$ quando $[H_{t},H_{t'}]=\mathbb{0}$
Vamos verificar que neste caso
\begin{equation}
U_{t} = e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}.
\end{equation}
Teremos
\begin{align}
i\hbar\partial_{t} U_{t} & = i\hbar\partial_{t} e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'} \\
                         & = i\hbar\partial_{t}\sum_{j=0}^{\infty}\frac{1}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j} \\
                         & = i\hbar\partial_{t}\left[\mathbb{I} + \sum_{j=1}^{\infty}\frac{1}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j}\right] \\
                         & = i\hbar\left[\mathbb{0} + \sum_{j=1}^{\infty}\frac{j}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j-1}\partial_{t}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)\right] \\
                         & = i\hbar\frac{-i}{\hbar}\partial_{t}\left(\int_{0}^{t}H_{t'}dt'\right)\sum_{j=1}^{\infty}\frac{j}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j-1},
\end{align}
No último passo usamos a comutatividade do Hamiltoniano em tempos diferentes. Seguindo, aplicando o teorema fundamental do cálculo e fazendo a mudança de variável $k=j-1$ obtemos
\begin{align}
i\hbar\partial_{t} U_{t} & = H_{t}\sum_{k=0}^{\infty}\frac{(k+1)}{(k+1)!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{k} \\
                         & = H_{t}\sum_{k=0}^{\infty}\frac{1}{k!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{k} \\ 
                         & = H_{t}e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'} \\
                         & = H_{t}U_{t}.
\end{align}

__Exercício:__ Verifique que se o Hamiltoniano puder ser escrito como $H_{t} = f(t)V$, com $f(t)$ sendo uma função escalar de $t$ e $V$ um operador Hermitiano independente do tempo, então $[H_{t},H_{t'}]=\mathbb{0}$. Nesse caso, use expansão espectral de $V=\sum_{j=1}^{\dim H}V_{j}|V_{j}\rangle\langle V_{j}|$ para verificar que $U_{t}=e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}$ satisfaz a equação de Schrödinger $i\hbar\partial_{t}U_{t}=H_{t}U_{t}$.
