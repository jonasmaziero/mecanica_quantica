# Momento angular (orbital)
Quantizando a expressão clássica
\begin{equation}
\vec{L} = \vec{X}\mathrm{x}\vec{P} = \sum_{j=1}^{3}\hat{e}_{j}(\vec{X}\mathrm{x}\vec{P})_{j} = \sum_{j=1}^{3}\hat{e}_{j}\sum_{k,l=1}^{3}\varepsilon_{jkl}X_{k}P_{l} = \sum_{j=1}^{3}\hat{e}_{j}L_{j}.
\end{equation}

## Relação de comutação para as componentes do momento angular (orbital)
Usando novamente $[A,BC]=B[A,C]+[A,B]C$ e $[AB,C] = A[B,C]+[A,C]B$, teremos
\begin{align}
[L_{j},L_{k}] & = \left[\sum_{l,m=1}^{3}\varepsilon_{jlm}X_{l}P_{m} , \sum_{p,q=1}^{3}\varepsilon_{kpq}X_{p}P_{q} \right] = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left[X_{l}P_{m} , X_{p}P_{q} \right] \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(X_{l}[P_{m} , X_{p}P_{q}] + [X_{l} , X_{p}P_{q}]P_{m} \right) \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(X_{l}X_{p}[P_{m},P_{q}] + X_{l}[P_{m},X_{p}]P_{q} + X_{p}[X_{l},P_{q}]P_{m} + [X_{l} , X_{p}]P_{q}P_{m} \right) \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(\mathbb{0} + X_{l}(-i\hbar\mathbb{1}\delta_{mp})P_{q} + X_{p}(i\hbar\mathbb{1}\delta_{lq})P_{m} + \mathbb{0} \right) \\
              & = -i\hbar\sum_{l,m,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kmq}X_{l}P_{q} + i\hbar\sum_{l,m,p=1}^{3}\varepsilon_{jlm}\varepsilon_{kpl}X_{p}P_{m} \\
\end{align}

Podemos verificar que
\begin{equation}
\sum_{j=1}^{3}\varepsilon_{jkl}\varepsilon_{jmn} = \delta_{km}\delta_{ln} - \delta_{kn}\delta_{lm}.
\end{equation}

Obtemos assim
\begin{align}
[L_{j},L_{k}] & = i\hbar\sum_{l,q=1}^{3}\sum_{m=1}^{3}\varepsilon_{mjl}\varepsilon_{mkq}X_{l}P_{q} - i\hbar\sum_{m,p=1}^{3}\sum_{l=1}^{3}\varepsilon_{ljm}\varepsilon_{lkp}X_{p}P_{m} \\
              & = i\hbar\sum_{l,q=1}^{3}(\delta_{jk}\delta_{lq}-\delta_{jq}\delta_{lk})X_{l}P_{q} - i\hbar\sum_{m,p=1}^{3}(\delta_{jk}\delta_{mp}-\delta_{jp}\delta_{mk})X_{p}P_{m}.
\end{align}
Como para $j=k$ temos $[L_{j},L_{k}]=\mathbb{O}$, vamos considerar a relação acima com $j\ne k\Rightarrow \delta_{jk}=0$. Assim
\begin{align}
[L_{j},L_{k}] & = -i\hbar\sum_{l,q=1}^{3}\delta_{jq}\delta_{lk}X_{l}P_{q} + i\hbar\sum_{m,p=1}^{3}\delta_{jp}\delta_{mk}X_{p}P_{m} \\
              & = i\hbar(-X_{k}P_{j} + X_{j}P_{k}).
\end{align}

Substituindo os índices,
\begin{align}
[L_{1},L_{2}] & = i\hbar(X_{1}P_{2}-X_{2}P_{1}) = i\hbar\sum_{j,k}\varepsilon_{3jk}X_{j}P_{k} = i\hbar L_{3} =  i\hbar\varepsilon_{123}L_{3}, \\
[L_{2},L_{3}] & = i\hbar(X_{2}P_{3}-X_{3}P_{2}) = i\hbar\sum_{j,k}\varepsilon_{1jk}X_{j}P_{k} = i\hbar L_{1} = i\hbar\varepsilon_{231}L_{1}, \\
[L_{1},L_{3}] & = i\hbar(X_{1}P_{3}-X_{3}P_{1}) = -i\hbar\sum_{j,k}\varepsilon_{2jk}X_{j}P_{k} = -i\hbar L_{2} = i\hbar\varepsilon_{132}L_{2}. \\
\end{align}
Finalmente a importante relação de comutação para as componentes do momento angular escrita em uma forma geral:
\begin{equation}
[L_{j},L_{k}] = i\hbar\varepsilon_{jkl}L_{l} = i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l}.
\end{equation}

## Autovalores do momento angular
Aqui, consideremos qualquer operador $\vec{L}$ cujas componentes satisfazem a relação de comutação: $[L_{j},L_{k}] = i\hbar\varepsilon_{jkl}L_{l}$. O método de diagonalização que usaremos aqui é semelhante àquele utilizado na solução do oscilador harmônico. Primeiramente verifiquemos que o quadrado do momento angular comuta com todas as suas componentes:
\begin{align}
[L^{2},L_{k}] & = \left[\sum_{j=1}^{3}L_{j}^{2},L_{k}\right] = \sum_{j=1}^{3}[L_{j}L_{j},L_{k}] = \sum_{j=1}^{3}(L_{j}[L_{j},L_{k}] + [L_{j},L_{k}]L_{j}) \\
              & = \sum_{j=1}^{3}(L_{j}i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l} + i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l}L_{j}) = i\hbar\sum_{j,l=1}^{3}\varepsilon_{jkl}(L_{j}L_{l} + L_{l}L_{j}) \\
              & = i\hbar\sum_{j,l=1}^{3}\varepsilon_{jkl}\{L_{j},L_{l}\}.
\end{align}
Considerando as componentes explicitamente,
\begin{align}
[L^{2},L_{1}] & = i\hbar(\varepsilon_{213}\{L_{2},L_{3}\} + \varepsilon_{312}\{L_{3},L_{2}\}) = i\hbar\{L_{2},L_{3}\}(\varepsilon_{213} + \varepsilon_{312}) \\ 
              & = i\hbar\{L_{2},L_{3}\}(\varepsilon_{213} - \varepsilon_{312}) = \mathbb{O},
\end{align}
onde usamos $\{L_{j},L_{k}\}=\{L_{k},L_{j}\}$. <br>
__Exercício:__ Faça a verificação análoga para $k=1$ e $k=2$. <br>
Com isso vemos que
\begin{equation}
[L^{2},L_{k}] = \mathbb{O}  \text{ para } k=1,2,3.
\end{equation}

Seguindo, consideremos que $L^{2}$ compartilha a base de autovetores $|l,m\rangle$ com $L_{3}$ e
\begin{align}
L^{2}|l,m\rangle & = \lambda|l,m\rangle \\
L_{3}|l,m\rangle & = \hbar m|l,m\rangle.
\end{align}
Vamos verificar que $\lambda = \hbar^{2}l(l+1)$ com $l=n/2$ para $n\in\mathbb{N}$ e $m=-l,-l+1,\cdots,-1,0,1\cdots,l-1,l$. Para isso precisaremos dos _operadores escada_:
\begin{equation}
L_{\pm} := L_{1} \pm iL_{2}.
\end{equation}
Como
\begin{equation}
[L^{2},L_{\pm}] = [L^{2},L_{1}] \pm i[L^{2},L_{2}] = \mathbb{O}, 
\end{equation}
vemos que os vetores $L_{\pm}|l,m\rangle$ também são autovetores de $L^{2}$ correspondentes aos mesmos autovalores $\lambda$ que $|l,m\rangle$. Por outro lado,
\begin{align}
[L_{3},L_{\pm}] & = [L_{3},L_{1}] \pm i[L_{3},L_{2}] = i\hbar\varepsilon_{312}L_{2} \pm i i\hbar\varepsilon_{321}L_{1}  = i\hbar L_{2} \pm \hbar L_{1} = \pm\hbar(L_{1} \pm iL_{2}) \\
& = \pm\hbar L_{\pm}
\end{align}
e
\begin{align}
L_{3}(L_{\pm}|l,m\rangle) & = (L_{3}L_{\pm})|l,m\rangle = (\pm\hbar L_{\pm} + L_{\pm}L_{3})|l,m\rangle = \pm\hbar L_{\pm}|l,m\rangle + L_{\pm}L_{3}|l,m\rangle = \pm\hbar L_{\pm}|l,m\rangle + L_{\pm}\hbar m|l,m\rangle \\
                          & = \hbar( m \pm 1)(L_{\pm}|l,m\rangle).
\end{align}

Ou seja, os vetores $L_{\pm}|l,m\rangle$ também são autovetores de $L_{3}$, mas com autovalores aumentados ou diminuídos de $\hbar$ em relação àquele associado com $|l,m\rangle$. Agora, como $L^{2} = L_{1}^{2}+L_{2}^{2}+L_{3}^{2}$, não poderemos aumentar nem diminuir indefinidamente os autovalores de $L_{3}$ através da aplicação dos operadores escada. Por isso, para os valores definidos por $m=\pm l$ a ação dos operadores escada nos retorna o vetor nulo:
\begin{equation}
L_{\pm}|l,m =\pm l\rangle := |\oslash\rangle.
\end{equation}

_OBS._ Note que como $m_{\min}=-l$ e $m_{\max}=l$ e $m$ aumenta ou diminiu de 1 em 1, para um dado $l$ teremos $2l+1$ possíveis valores de $m$. Portanto o espaço vetorial usado para definir o conjunto de estados de momento angular é $\mathbb{C}^{2l+1}$.

Continuando, consideremos
\begin{align}
L_{\pm}L_{\mp} & = (L_{1} \pm iL_{2})(L_{1} \mp iL_{2}) = L_{1}^{2} \mp iL_{1}L_{2} \pm iL_{2}L_{1} + L_{2}^{2}  = L_{1}^{2} + L_{2}^{2} + L_{3}^{2} - L_{3}^{2} \mp i[L_{1},L_{2}]  = L^{2} - L_{3}^{2} \mp ii\hbar\varepsilon_{123}L_{3} \\
               & = L^{2} - L_{3}^{2} \pm\hbar L_{3}.
\end{align}
Então, podemos considerar
\begin{align}
L^{2}|l,l\rangle & = (L_{-}L_{+} + L_{3}^{2} + \hbar L_{3})|l,l\rangle  = L_{-}L_{+}|l,l\rangle + L_{3}^{2}|l,l\rangle + \hbar L_{3}|l,l\rangle  = |\oslash\rangle + \hbar^{2}l^{2}|l,l\rangle + \hbar\hbar l|l,l\rangle \\
                 & = \hbar^{2}l(l+1)|l,l\rangle  = \lambda|l,l\rangle.
\end{align}

Para obter o __domínio de $l$__, consideremos que os autovalores de $L_{3}$ são, por definição, $\hbar m$ e que vimos que se começamos no estado $|l,m=-l\rangle$ e aplicamos $L_{+}$ $n$ vezes, aumentaremos o valor de $m$ de uma unidade em cada aplicação de $L_{+}$, até obtermos o estado $|l,m=l\rangle$. Como $n\in\mathbb{N}$ teremos que $l=-l+n\Rightarrow 2l=n \Rightarrow l = n/2 = 0, 1/2, 1, 3/2, \cdots$.

### Ação dos operadores escada na base $|l,m\rangle$
Como os operadores escada aumentam ou diminuem o número quântico $m$ de uma unidade, teremos e.g.
\begin{equation}
L_{+}|l,m\rangle \propto |l,m+1\rangle =: c|l,m+1\rangle,
\end{equation}
onde introduzimos a constate $c$ para transformar a proporcionalidade em igualdade. Olhemos agora para o quadrado da norma desse vetor
\begin{align}
\langle l,m|L_{+}^{\dagger}L_{+}|l,m\rangle & = \langle l,m|L_{-}L_{+}|l,m\rangle = \langle l,m|(L^{2} - L_{3}^{2} -\hbar L_{3})|l,m\rangle \\
& = \langle l,m|\left(\hbar^{2}l(l+1)-\hbar^{2}m^{2}-\hbar^{2}m\right)|l,m\rangle = \left(\hbar^{2}l(l+1)-\hbar^{2}m^{2}-\hbar^{2}m\right)\langle l,m|l,m\rangle \\
& = \hbar^{2}[l(l+1) - m(m+1)]  \\
& = \langle l,m+1|c^{*}c|l,m+1\rangle = |c|^{2}.
\end{align}
Usando $c\in\mathbb{R}$ obtemos
\begin{equation}
L_{+}|l,m\rangle = \hbar\sqrt{l(l+1)-m(m+1)}|l,m+1\rangle.
\end{equation}

# Soma de momento angular
Consideremos os momentos angulares $\vec{J}_{s}$ com $s=a,b$. Ou seja, $[J_{j}^{s},J_{k}^{s}]=i\hbar\varepsilon_{jkl}J_{l}^{s}$ para $j,k=1,2,3$. Para esses dois graus de liberdade, a soma desses momentos angulares é representada por um operador Hermitiano que atua no espaço composto $\mathcal{H}_{a}\otimes\mathcal{H}_{b}$ e que é escrito como segue:
\begin{equation}
\vec{J} = \vec{J}_{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes\vec{J}_{b} = \sum_{i=1}^{3}\hat{e}_{i}\left(J_{i}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{i}^{b}\right).
\end{equation}

Para os conjuntos de observáveis
\begin{equation}
\{J_{a}^{2},J_{3}^{a},J_{b}^{2},J_{3}^{b}\} \text{ e } \{J_{a}^{2},J_{b}^{2},J^{2},J_{3}\},
\end{equation}
verificaremos na sequência que os observáveis dentro de cada um dos conjuntos formam um conjunto de observáveis mutuamente compatíveis. Bom, já temos que $[J_{s}^{2},J_{3}^{s}]=\mathbb{O}$. Para os cálculos que seguem, primeiro usamos $(A\otimes B)(C\otimes D) = AC\otimes BD$, que vale para quaisquer operadores $A,C:\mathcal{H}_{a}\rightarrow\mathcal{H}_{a}$ e $B,D:\mathcal{H}_{b}\rightarrow\mathcal{H}_{b}$, para ver que 
\begin{align}
[A\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes B] & = (A\otimes\mathbb{I}_{b})(\mathbb{I}_{a}\otimes B) - (\mathbb{I}_{a}\otimes B)(A\otimes\mathbb{I}_{b}) = (A\mathbb{I}_{a}\otimes\mathbb{I}_{b}B) - (\mathbb{I}_{a}A\otimes B\mathbb{I}_{b}) \\
& = A\otimes B - A\otimes B = \mathbb{0}.
\end{align}
Esse resultado pode ser aplicado diretamente a
\begin{align}
& [J_{a}^{2},J_{b}^{2}] = [J_{a}^{2}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{b}^{2}] = \mathbb{O}, \\
& [J_{a}^{2},J_{3}^{b}] = [J_{a}^{2}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{3}^{b}] = \mathbb{O}, \\
& [J_{b}^{2},J_{3}^{a}] = [\mathbb{I}_{a}\otimes J_{b}^{2},J_{3}^{a}\otimes\mathbb{I}_{b}] = \mathbb{O}, \\
& [J_{3}^{a},J_{3}^{b}] = [J_{3}^{a}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{3}^{b}] = \mathbb{O}.
\end{align}

Agora vamos para o segundo conjunto de operadores. Aqui começamos verificando que $\vec{J}$ é de fato um momento angular:
\begin{align}
[J_{j},J_{k}] & = [J_{j}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{j}^{b}, J_{k}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{k}^{b}] \\
              & = [J_{j}^{a}\otimes\mathbb{I}_{b},J_{k}^{a}\otimes\mathbb{I}_{b}] + [J_{j}^{a}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{k}^{b}] + [\mathbb{I}_{a}\otimes J_{j}^{b},J_{k}^{a}\otimes\mathbb{I}_{b}] + [\mathbb{I}_{a}\otimes J_{j}^{b},\mathbb{I}_{a}\otimes J_{k}^{b}] \\
              & = [J_{j}^{a},J_{k}^{a}]\otimes\mathbb{I}_{b} + \mathbb{0} + \mathbb{0} + \mathbb{I}_{a}\otimes[J_{j}^{b},J_{k}^{b}] = i\hbar\varepsilon_{jkl}J_{l}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes i\hbar\varepsilon_{jkl}J_{l}^{b} \\
              & = i\hbar\varepsilon_{jkl}\left(J_{l}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{l}^{b}\right)  = i\hbar\varepsilon_{jkl}J_{l}.
\end{align}
Pela mesma demonstração feita no caso do momento angular orbital, temos aqui que
\begin{equation}
[J^{2},J_{j}] = \mathbb{O}
\end{equation}
para $j=1,2,3$. Desse resultado segue que $[J^{2},J_{3}]=\mathbb{O}$. Seguindo, notemos que (__exercício__)
\begin{equation}
J^{2} = \vec{J}\cdot\vec{J} =  J_{a}^{2}\otimes\mathbb{I}_{b} +  \mathbb{I}_{a}\otimes J_{b}^{2} + 2\sum_{j=1}^{3}J_{j}^{a}\otimes J_{j}^{b}. \\
\end{equation}

Usando esse resultado, vemos que

[J^{2},J_{a}^{2}] & = [J^{2},J_{a}^{2}\otimes\mathbb{I}_{b}] \\
                  & = [J_{a}^{2}\otimes\mathbb{I}_{b},J_{a}^{2}\otimes\mathbb{I}_{b}] + [\mathbb{I}_{a}\otimes J_{b}^{2},J_{a}^{2}\otimes\mathbb{I}_{b}] + [2\sum_{j=1}^{3}J_{j}^{a}\otimes J_{j}^{b},J_{a}^{2}\otimes\mathbb{I}_{b}] \\

\end{align}

Nos falta ainda olhar para

[J_{a}^{2},J_{3}] & = [J_{a}^{2}\otimes\mathbb{I}_{b},J_{3}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{3}^{b}] = [J_{a}^{2}\otimes\mathbb{I}_{b},J_{3}^{a}\otimes\mathbb{I}_{b}] + [J_{a}^{2}\otimes\mathbb{I}_{b},\mathbb{I}_{a}\otimes J_{3}^{b}] \\
                  & = \mathbb{O}.
\end{align}
e para (__exercício__)

Vamos denotar os autovetores comuns de $\{J_{a}^{2},J_{3}^{a},J_{b}^{2},J_{3}^{b}\}$ por $|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle$ e para os autovetores comuns de $\{J_{a}^{2},J_{b}^{2},J^{2},J_{3}\}$ usaremos $|j_{a},j_{b},j,m\rangle$. Assim
\begin{align}
& J_{a}^{2}\otimes\mathbb{I}_{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle = \hbar^{2}j_{a}(j_{a}+1)|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle, \\
& J_{3}^{a}\otimes\mathbb{I}_{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle = \hbar m_{a}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle, \\
& \mathbb{I}_{a}\otimes J_{b}^{2}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle = \hbar^{2}j_{b}(j_{b}+1)|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle, \\
& \mathbb{I}_{a}\otimes J_{3}^{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle = \hbar m_{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle
\end{align}
e
\begin{align}
& J_{a}^{2}\otimes\mathbb{I}_{b}|j_{a},j_{b},j,m\rangle = \hbar^{2}j_{a}(j_{a}+1)|j_{a},j_{b},j,m\rangle, \\
& \mathbb{I}_{a}\otimes J_{b}^{2}|j_{a},j_{b},j,m\rangle = \hbar^{2}j_{b}(j_{b}+1)|j_{a},j_{b},j,m\rangle, \\
& J^{2}|j_{a},j_{b},j,m\rangle = \hbar^{2}j(j+1)|j_{a},j_{b},j,m\rangle, \\
& J_{3}|j_{a},j_{b},j,m\rangle = \hbar m|j_{a},j_{b},j,m\rangle.
\end{align}

Notemos ainda que
\begin{align}
J_{3}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle & = J_{3}^{a}\otimes\mathbb{I}_{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle + \mathbb{I}_{a}\otimes J_{3}^{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle \\
               & = \hbar m_{a}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle + \hbar m_{b}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle \\
               & = \hbar (m_{a}+m_{b})|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle.
\end{align}

Ou seja, $|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle$ também é autovetor de $J_{3}$ com autovalor $\hbar(m_{a}+m_{b})$. No entanto, as componentes dos momentos angulares individuias não comutam com o momento angular total. Por exemplo:
\begin{align}
[J^{2},J_{3}^{a}] & = \sum_{j=1}^{3}[J_{j}^{2},J_{3}^{a}\otimes\mathbb{I}_{b}] = \sum_{j=1}^{3}\left(J_{j}[J_{j},J_{3}^{a}\otimes\mathbb{I}_{b}] + [J_{j},J_{3}^{a}\otimes\mathbb{I}_{b}]J_{j} \right) \\
                  & = \sum_{j=1}^{3}\left(J_{j}[J_{j}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{j}^{b},J_{3}^{a}\otimes\mathbb{I}_{b}] + [J_{j}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{j}^{b},J_{3}^{a}\otimes\mathbb{I}_{b}]J_{j} \right) \\
                & = \sum_{j=1}^{3}\left(J_{j}([J_{j}^{a},J_{3}^{a}]\otimes\mathbb{I}_{b}+\mathbb{O}) + ([J_{j}^{a},J_{3}^{a}]\otimes\mathbb{I}_{b}+\mathbb{O})J_{j} \right) \\
                & = \sum_{j=1}^{3}i\hbar \varepsilon_{j3k}\left(J_{j}J_{k}^{a}\otimes\mathbb{I}_{b} + J_{k}^{a}\otimes\mathbb{I}_{b}J_{j} \right) \\
                & = i\hbar \varepsilon_{13k}\left(J_{1}J_{k}^{a}\otimes\mathbb{I}_{b} + J_{k}^{a}\otimes\mathbb{I}_{b}J_{1} \right) + i\hbar \varepsilon_{23k}\left(J_{2}J_{k}^{a}\otimes\mathbb{I}_{b} + J_{k}^{a}\otimes\mathbb{I}_{b}J_{2} \right) \\
                & = i\hbar \varepsilon_{132}\left(J_{1}J_{2}^{a}\otimes\mathbb{I}_{b} + J_{2}^{a}\otimes\mathbb{I}_{b}J_{1} \right) + i\hbar \varepsilon_{231}\left(J_{2}J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes\mathbb{I}_{b}J_{2} \right) \\
                & = i\hbar\left(J_{2}J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes\mathbb{I}_{b}J_{2} - J_{1}J_{2}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes\mathbb{I}_{b}J_{1} \right) \\
                & = i\hbar(J_{2}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{2}^{b})J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes\mathbb{I}_{b}(J_{2}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{2}^{b}) \\
                & - (J_{1}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{1}^{b})J_{2}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes\mathbb{I}_{b}(J_{1}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{1}^{b})  \\
                & = i\hbar(J_{2}^{a}J_{1}^{a}\otimes\mathbb{I}_{b} + J_{1}^{a}\otimes J_{2}^{a} + J_{1}^{a}J_{2}^{a}\otimes\mathbb{I}_{b}  + J_{1}^{a}\otimes J_{2}^{a} \\
                & - J_{1}^{a}J_{2}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes J_{1}^{a} - J_{2}^{a}J_{1}^{a}\otimes\mathbb{I}_{b} - J_{2}^{a}\otimes J_{1}^{a}) \\
                & = 2i\hbar(J_{1}^{a}\otimes J_{2}^{a} - J_{2}^{a}\otimes J_{1}^{a}) \\
                & \ne \mathbb{O}.                
\end{align}
Analogamente (__exercício__)
\begin{equation}
[J^{2},J_{3}^{b}] \ne \mathbb{O}.
\end{equation}

Estes últimos resultados nos mostram que, para os momentos angulares individuais fixos (i.e., para os valores dos números quânticos $j_{s}$ fixos), estados $|j_{a},j_{b},j,m\rangle$ com momento angular total fixo devem corresponder a superposições de dois ou mais estados $|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle$ com valores fixos das componentes $z$ dos momentos angulares individuais. Para escrevermos essa superposição explicitamente, vamos utilizar
\begin{equation}
\sum_{m_{s}=-j_{s}}^{j_{s}}|j_{s},m_{s}\rangle\langle j_{s},m_{s}| = \mathbb{I}_{j_{s}}
\end{equation}
para escrever o operador identidade no subespaço $\mathcal{H}_{j_{a}}\otimes\mathcal{H}_{j_{b}}\subset\mathcal{H}$ gerado por estados com $\{j_{s}\}_{s=a,b}$ constantes. Segue assim que
\begin{align}
|j_{a},j_{b},j,m\rangle & = \mathbb{I}_{j_{a}}\otimes\mathbb{I}_{j_{b}}|j_{a},j_{b},j,m\rangle \\
                        & = \left(\sum_{m_{a}=-j_{a}}^{j_{a}}|j_{a},m_{a}\rangle\langle j_{a},m_{a}|\otimes\sum_{m_{b}=-j_{b}}^{j_{b}}|j_{b},m_{b}\rangle\langle j_{b},m_{b}|\right)|j_{a},j_{b},j,m\rangle \\
                        & = \sum_{m_{a}=-j_{a}}^{j_{a}}\sum_{m_{b}=-j_{b}}^{j_{b}}(|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle)(\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)|j_{a},j_{b},j,m\rangle \\
                        & =: \sum_{m_{a}=-j_{a}}^{j_{a}}\sum_{m_{b}=-j_{b}}^{j_{b}} C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|j_{a},m_{a}\rangle\otimes|j_{b},m_{b}\rangle,
\end{align}
onde usamos novamente $AB\otimes CD=(A\otimes C)(B\otimes D)$ e definimos os chamados _coeficientes de Clebsch-Gordan_ como:
\begin{equation}
C_{m_{a}m_{b}m}^{j_{a}j_{b}j} := (\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)|j_{a},j_{b},j,m\rangle.
\end{equation}

Existem tabelas desses coeficientes, cuja obtenção usa restrições algébricas e simetrias. Por exemplo, uma condição necessária para que os coeficientes sejam não nulos é que $m=m_{a}+m_{b}$. Para verificar essa afirmação consideremos (atua $J_{3}$ na esquerda e $J_{3}^{s}$ na direita)
\begin{align}
& (\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)(J_{3} - J_{3}^{a}\otimes\mathbb{I}_{b} - \mathbb{I}_{a}\otimes J_{3}^{b})|j_{a},j_{b},j,m\rangle = (\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)\mathbb{O}|j_{a},j_{b},j,m\rangle = 0, \\
& \hbar(m-m_{a}-m_{b})(\langle j_{a},m_{a}|\otimes\langle j_{b},m_{b}|)|j_{a},j_{b},j,m\rangle = 0, \\
& \hbar(m-m_{a}-m_{b})C_{m_{a}m_{b}m}^{j_{a}j_{b}j} = 0.
\end{align}

Como a base $|j_{s},m_{s}\rangle$ possiu $2j_{s}+1$ vetores, temos a dimensão $\dim \mathcal{H}_{j_{a}}\otimes\mathcal{H}_{j_{b}} = (2j_{a}+1)(2j_{b}+1)$, que é igual ao número de estados da base $\{|j_{a},j_{b},j,m\rangle\}$. Agora, fixados os valores de $j_{s}$, conheceremos também os possíveis valores de $m_{s}$, e consequentemente os possíveis valores de $m$. Para obter os possíveis valores de $j$, notemos que o maior momento angular total é obtido se $\vec{J}_{a}$ e $\vec{J}_{b}$ forem "paralelos", ou seja, se tiverem e.g. os valores máximos das componentes $z$ dos momentos angulares individuais, i.e., se $\{m_{s}=j_{s}\}_{s=a,b}$. Assim $m=m_{a}+m_{b}=j_{a}+j_{b}=j$. Por outro lado, o menor valor possível de $j$ é obtido se $\vec{J}_{a}$ e $\vec{J}_{b}$ forem "antiparalelos", ou seja, se e.g. $m_{a}=j_{a}$ e $m_{b}=-j_{b}$ ou $m_{a}=-j_{a}$ e $m_{b}=j_{b}$. Para ter o mínimo entre $m=j_{a}-j_{b}$ e $m=j_{b}-j_{a}$ usamos como o menor valor de $j$ o módulo $|j_{a}-j_{b}|$. Com isso segue que $j\in\mathbb{N}$ com
\begin{equation}
|j_{a}-j_{b}| < j < j_{a}+j_{b}.
\end{equation}

## __Exemplo:__ Soma de dois spins 1/2
Como um primeiro exemplo, consideremos que $j_{s}=1/2$ para $s=a,b$. Nesse caso $m_{s}=-1/2,1/2$, $m=-1,0,1$ e $j=0,1$. Claro, para $j=0$ devemos ter $m=0$. Vamos determinar os coeficientes de Clebsch-Gordan nesse caso. Por conveniência, não escreveremos $j_{s}$ nos labels dos vetores de estado. Usando a restrição $m=m_{a}+m_{b}$ e começando com o estado de maior componente $z$ do momento angular total, vemos que
\begin{align}
|j=1,m=1\rangle & = \sum_{m_{a}=-1/2}^{1/2}\sum_{m_{b}=-1/2}^{1/2}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle \\
                & = C_{-1/2,-1/2,1}^{1/2,1/2,1}|-1/2\rangle\otimes|-1/2\rangle + C_{-1/2,1/2,1}^{1/2,1/2,1}|-1/2\rangle\otimes|1/2\rangle + C_{1/2,-1/2,1}^{1/2,1/2,1}|1/2\rangle\otimes|-1/2\rangle + C_{1/2,1/2,1}^{1/2,1/2,1}|1/2\rangle\otimes|1/2\rangle \\
                & = 0 + 0 + 0 + C_{1/2,1/2,1}^{1/2,1/2,1}|1/2\rangle\otimes|1/2\rangle \\
                & = |m_{a}=1/2\rangle\otimes|m_{b}=1/2\rangle.
\end{align}

Seguindo, aplicamos o operador de abaixamento $J_{-} = J_{1}-iJ_{2}$ nesse estado, lembrando que $J_{-}|j,m\rangle=\hbar\sqrt{j(j+1)-m(m-1)}|j,m-1\rangle$,
para obter $J_{-}|j=1,m=1\rangle=\hbar\sqrt{1(1+1)-1(1-1)}|1,1-1\rangle=\hbar\sqrt{2}|1,0\rangle$. Assim
\begin{align}
|j=1,m=0\rangle & = (\hbar^{-1}/\sqrt{2})J_{-}|j=1,m=1\rangle \\
                & = (\hbar^{-1}/\sqrt{2})(J_{-}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{-}^{b})|1/2\rangle\otimes|1/2\rangle \\
                & = (\hbar^{-1}/\sqrt{2})((J_{-}^{a}|1/2\rangle)\otimes|1/2\rangle + |1/2\rangle\otimes(J_{-}^{b}|1/2\rangle)) \\
                & = (\hbar^{-1}/\sqrt{2})((\hbar\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}-1)}|1/2-1\rangle)\otimes|1/2\rangle + |1/2\rangle\otimes(\hbar\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}-1)}|1/2-1\rangle)) \\
                & = (1/\sqrt{2})(\sqrt{1/2(1/2+1)-(1/2)(1/2-1)}|-1/2\rangle\otimes|1/2\rangle + \sqrt{(1/2)(1/2+1)-(1/2)(1/2-1)}|1/2\rangle\otimes|-1/2\rangle) \\
                & = (1/\sqrt{2})(\sqrt{3/4+1/4}|-1/2\rangle\otimes|1/2\rangle + \sqrt{3/4+1/4}|1/2\rangle\otimes|-1/2\rangle) \\
                & = (1/\sqrt{2})\left(|m_{a}=-1/2\rangle\otimes|m_{b}=1/2\rangle + |m_{a}=1/2\rangle\otimes|m_{b}=-1/2\rangle\right). \\
\end{align}

Ainda, $J_{-}|j=1,m=0\rangle=\hbar\sqrt{1(1+1)-0(0-1)}|1,0-1\rangle=\hbar\sqrt{2}|1,-1\rangle$. Assim
\begin{align}
|j=1,m=-1\rangle & = (\hbar^{-1}/\sqrt{2})J_{-}|j=1,m=0\rangle \\
                 & = (\hbar^{-1}/2)(J_{-}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{-}^{b})(|-1/2\rangle\otimes|1/2\rangle + |1/2\rangle\otimes|-1/2\rangle) \\
                 & = (\hbar^{-1}/2)(J_{-}^{a}\otimes\mathbb{I}_{b}|-1/2\rangle\otimes|1/2\rangle + J_{-}^{a}\otimes\mathbb{I}_{b}|1/2\rangle\otimes|-1/2\rangle + \mathbb{I}_{a}\otimes J_{-}^{b}|-1/2\rangle\otimes|1/2\rangle + \mathbb{I}_{a}\otimes J_{-}^{b}|1/2\rangle\otimes|-1/2\rangle) \\
                 & = (1/2)(|\oslash\rangle\otimes|1/2\rangle + \sqrt{1}|-1/2\rangle\otimes|-1/2\rangle + \sqrt{1}|-1/2\rangle\otimes|-1/2\rangle + |1/2\rangle\otimes|\oslash\rangle) \\
                 & = |m_{a}=-1/2\rangle\otimes|m_{b}=-1/2\rangle.
\end{align}

Construíremos o estado que falta da base como o primeiro
\begin{align}
|j=0,m=0\rangle & = \sum_{m_{a}=-1/2}^{1/2}\sum_{m_{b}=-1/2}^{1/2}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle \\
                & = C_{-1/2,-1/2,0}^{1/2,1/2,0}|-1/2\rangle\otimes|-1/2\rangle + C_{-1/2,1/2,0}^{1/2,1/2,0}|-1/2\rangle\otimes|1/2\rangle + C_{1/2,-1/2,0}^{1/2,1/2,0}|1/2\rangle\otimes|-1/2\rangle + C_{1/2,1/2,0}^{1/2,1/2,0}|1/2\rangle\otimes|1/2\rangle \\
                & = 0 + C_{-1/2,1/2,0}^{1/2,1/2,0}|-1/2\rangle\otimes|1/2\rangle + C_{1/2,-1/2,0}^{1/2,1/2,0}|1/2\rangle\otimes|-1/2\rangle + 0 \\
                & = (1/\sqrt{2})\left(|m_{a}=-1/2\rangle\otimes|m_{b}=1/2\rangle - |m_{a}=1/2\rangle\otimes|m_{b}=-1/2\rangle\right),
\end{align}
com a escolha dos coeficientes de Clebsch-Gordan feita de forma que o estado seja normalizado e ortogonal aos outros estados da base.

### Relação de recursão para os coeficientes de Clebsch-Gordan
Vamos considerar $j_{a},j_{b},j$ fixos. De $|j,m\rangle=\sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle$ teremos que
\begin{align}
J_{\pm}|j,m\rangle & = \hbar\sqrt{j(j+1)-m(m\pm1)}|j,m\pm1\rangle \\
        & = (J_{\pm}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{\pm}^{b})|j,m\rangle \\
        & = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\left(J_{\pm}^{a}|m_{a}\rangle\otimes|m_{b}\rangle + |m_{a}\rangle\otimes J_{\pm}^{b}|m_{b}\rangle\right) \\
        & = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\hbar\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\pm1)}|m_{a}\pm1\rangle\otimes|m_{b}\rangle \\
         & + \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\hbar\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\pm1)}|m_{a}\rangle\otimes|m_{b}\pm1\rangle. 
\end{align}
Tomando o produto interno de $|m_{a}'\rangle\otimes|m_{b}'\rangle$ com esta última equação dividida por $\hbar$ vem que
\begin{align}
& \sqrt{j(j+1)-m(m\pm1)}\langle m_{a}'|\otimes\langle m_{b}'|j,m\pm1\rangle \\
& = \sqrt{j(j+1)-m(m\pm1)}C_{m_{a}'m_{b}'(m\pm1)}^{j_{a}j_{b}j} \\
& = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\pm1)}(\langle m_{a}'|\otimes\langle m_{b}'|)(|m_{a}\pm1\rangle\otimes|m_{b}\rangle) \\ 
& + \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\pm1)}(\langle m_{a}'|\otimes\langle m_{b}'|)(|m_{a}\rangle\otimes|m_{b}\pm1\rangle) \\
& = \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\pm1)}\delta_{m_{a}',m_{a}\pm1}\delta_{m_{b}',m_{b}} \\ 
& + \sum_{m_{a},m_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\pm1)}\delta_{m_{a}',m_{a}}\delta_{m_{b}',m_{b}\pm1} \\
& = C_{(m_{a}'\mp1)m_{b}'m}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}'(m_{a}'\mp1)} \\
& + C_{m_{a}'(m_{b}'\mp1)m}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}'(m_{b}'\mp1)}.
\end{align}
Usando $m'=m\pm1$ e trocando os números quânticos com linha pelos sem obtemos
\begin{align}
\sqrt{j(j+1)-m(m\mp1)}C_{m_{a}m_{b}m}^{j_{a}j_{b}j} & = C_{(m_{a}\mp1)m_{b}(m\mp1)}^{j_{a}j_{b}j}\sqrt{j_{a}(j_{a}+1)-m_{a}(m_{a}\mp1)} \\
& + C_{m_{a}(m_{b}\mp1)(m\mp1)}^{j_{a}j_{b}j}\sqrt{j_{b}(j_{b}+1)-m_{b}(m_{b}\mp1)}.
\end{align}
