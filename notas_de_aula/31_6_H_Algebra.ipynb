# Átomos Hidrogenóides

## Solução pelo método algébrico

Vamos considerar o hamiltoniano para um potencial esfericamente simétrico escrito como
$$H = \frac{P^2}{2m} + \frac{\hbar^2 l(l+1)}{2mr^2} + U(r), \hspace{2cm} (\star)$$
em que $L^2$ compartilha outovetores com $H_0$ e $L^2$ tem autovalores iguais a $\hbar^2 l(l+1)$. Também usamos a notação
$$P=P_r.$$
Esse hamiltoniano pode descrever e.g. o átomo de Hidrogenio, o oscilador harmônico quântico tridimensional isotrópico e pode ser usado como uma aproximação para a força forte na aproximação de campo médio de partículas independentes no núcleo.

### Construção da álgebra

Para o momento angular, começamos com as equações de autovalores e autovetores 
\begin{align}
& L^2|\lambda,\mu\rangle = \lambda|\lambda,\mu\rangle,\\
& L_3|\lambda,\mu\rangle = \mu|\lambda,\mu\rangle.
\end{align}
Usando as relações de comutação canônicas (RCC)
\begin{align}
& [X_j,X_k]=0,\\
& [P_j,P_k]=0,\\
& [X_j,P_k]=i\hbar I\delta_{j,k},
\end{align}
para $j,k=1,2,3,$ obtemos as relações comutação que definem a álgebra do momento angular, ou a álgebra de Lie $so(3)$:
$$[L_j,L_k]=i\hbar\epsilon_{jkl}L_l$$
ou
\begin{align}
& [L_1,L_2]=i\hbar L_3, \\
& [L_2,L_3]=i\hbar L_1, \\
& [L_3,L_1]=i\hbar L_2.
\end{align}
Definindo os operadores escada
$$L_{\pm}= L_1 \pm iL_2$$
obtivemos
$$\lambda = \hbar^2 l(l+1),\ \mu=\hbar m$$
com
$$l=0,1,2,\cdots;\ m=-l,-l+1,\cdots,l.$$

Quanto à algebra do hamiltoniano acima, vamos começar usando as RCC para verificar que
$$[R,P^m] = i\hbar mP^{m-1},$$
em que $m\in\mathbb{N}.$ Para isso usaremos as identidades
\begin{align}
& [A,BC] = [A,B]C + B[A,C],\\
& [AB,C] = A[B,C] + [A,B]C
\end{align}
para obter
\begin{align}
[X,P] & = i\hbar I,\\
[X,P^2] & = P[X,P] + [X,P]P = P i\hbar I + i\hbar I P = i\hbar 2P, \\
[X,P^3] & = P[X,P^2] + [X,P]P^2 = P i\hbar 2P + i\hbar I P^2 = i\hbar 3P^2, \\
[X,P^4] & = P[X,P^3] + [X,P]P^3 = P i\hbar 3P^2 + i\hbar I P^3 = i\hbar 4P^3, \\
& \vdots \\
[X,P^m] & = i\hbar mP^{m-1}.
\end{align}

Da mesma forma, podemos verificar que
\begin{align}
[X,P] & = i\hbar I,\\
[X^2,P] & = X[X,P] + [X,P]X = X i\hbar I + i\hbar I X = i\hbar 2X, \\
[X^3,P] & = X[X^2,P] + [X,P]x^2 = X i\hbar 2X + i\hbar I X^2 = i\hbar 3X^2, \\
[X^4,P] & = X[X^3,P] + [X,P]X^3 = X i\hbar 3X^2 + i\hbar I X^3 = i\hbar 4X^3, \\
& \vdots \\
[X^m,P] & = i\hbar mX^{m-1}.
\end{align}

Considerando o hamiltoniano $(\star)$, vamos usar os produtos de Lie
$$\{R^a, R^b P, R^c P^2\},$$
com $a,b,c\in\mathbb{Z}$, pois $P^2$ é a maior potência de $P$ que aparece em $H$. Assim como a ocorre com as componentes do momento angular, queremos que esses três operadores fechem a álgebra, ou seja, que formem um conjunto fechado sob comutação. 

Usando as relações de comutação acima, e notando que temos efetivamente um problema unidimentsional com 
$$(X,P)\rightarrow (R,P)=(R,P_r),$$
vamos olhar para o comutador
\begin{align}
[R^a,R^b P] & = R^b[R^a, P] + [R^a,R^b]P \\
& = R^b i\hbar aR^{a-1} + 0 \\
& = i\hbar a R^{b+a-1}.
\end{align}

Se escolhemos
$$b=1,\ c=2-a$$
nosso conjunto de operadores vira
$$\{R^a, RP, R^{2-a}P^2\},$$
e teremos que
$$[R^a,R P] = i\hbar a R^{1+a-1} = i\hbar a R^a$$
e
\begin{align}
[R^a,R^c P^2] & = [R^a,R^{2-a} P^2]  = [R^a,R^{2-a} PP] \\
& = R^{2-a} P[R^a,P] + [R^a,R^{2-a} P]P \\
& = R^{2-a} P i\hbar a R^{a-1} + i\hbar a R^{2-a+a-1}P \\
& = i\hbar a R^{2-a}\big(R^{a-1} P -  i\hbar (a-1)R^{a-2}\big) + i\hbar a RP \\
& = i\hbar a RP + \hbar^2 a(a-1)R^0 + i\hbar a RP \\
& = 2i\hbar a RP + \hbar^2 a(a-1)I,
\end{align}
em que usamos
\begin{align}
& [R^{a-1},P]=R^{a-1} P - PR^{a-1} = i\hbar (a-1)R^{a-2}\\
& \therefore\ PR^{a-1} = R^{a-1} P -  i\hbar (a-1)R^{a-2}.
\end{align}

Também temos que
\begin{align}
[R^b P,R^c P^2] & = [R P,R^{2-a} P^2] \\
& = R^{2-a}[R P, P^2] + [R P,R^{2-a}] P^2 \\
& = R^{2-a}\big(R[P, P^2]+[R, P^2]P\big) + \big(R[P,R^{2-a}]+[R,R^{2-a}]P\big)P^2 \\
& = R^{2-a}\big(R 0+[R, P^2]P\big) + \big(R[P,R^{2-a}]+0P\big)P^2 \\
& = R^{2-a}\big(\big(P[R, P]+[R, P]P\big)P\big) - R i\hbar(2-a)R^{2-a-1}P^2 \\
& = 2i\hbar R^{2-a}P^2 - i\hbar(2-a)R^{2-a}P^2 \\
& = i\hbar aR^{2-a}P^2 \\
& = i\hbar R^c P^2.
\end{align}

Agora vamos fazer a seguinte mudança de variável
\begin{align}
& V_1 := R^a, \\
& V_2 := \frac{1}{a}\big(RP - i\hbar(a-1)I/2\big), \\
& V_3 := \frac{1}{a^2}R^{2-a}P^2.
\end{align}

Usando as relações de comutação obtidas acima, teremos que esses operadores satisfazem as seguintes relações de comutação
\begin{align}
[V_1,V_2] & = \big[R^a,\frac{1}{a}\big(RP - i\hbar(a-1)I/2\big)\big] = \frac{1}{a}\big(i\hbar aR^a + 0\big) \\
& = i\hbar V_1, \\
[V_2,V_3] & = \Big[\frac{1}{a}\big(RP - i\hbar(a-1)I/2\big), \frac{1}{a^2}R^{2-a}P^2\Big] \\
& = \frac{1}{a^3} \big(i\hbar aR^{2-a}P^2 + 0\big) = i\hbar\frac{1}{a^2}R^{2-a}P^2 \\
& = i\hbar V_3, \\
[V_1,V_3] & = \Big[R^a, \frac{1}{a^2}R^{2-a}P^2\Big] = \frac{1}{a^2}\big(2i\hbar a RP + \hbar^2 a(a-1)\big) \\
& = 2i\hbar\frac{1}{a}\big(RP -i\hbar(a-1)I/2\big) \\
& = 2i\hbar V_2.
\end{align}

Lembrando que $a\in\mathbb{Z}$, notemos que
$$R^{-a} = V_1^{-1}$$
e fazemos uma outra mudança de variável
\begin{align}
& T_1 := \frac{1}{2}\big(V_3 + \tau V_1^{-1} - V_1\big), \\
& T_2 := V_2, \\
& T_3 := \frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big), \\
\end{align}
em que $\tau$ é um operador constante qualquer que comuta com $V_1,\ V_2,\ V_3,$ i.e.,
$$[\tau,V_1]=[\tau,V_2]=[\tau,V_3]=0.$$
Notando que
\begin{align}
[V_2,V_1^{-1}] & = \Big[\frac{1}{a}\big(RP - i\hbar(a-1)I/2\big),R^{-a}\Big] = -\frac{1}{a}\big[R^{-a},RP\big] = -\frac{1}{a}\big[R^{a'},RP\big] \\
& = -\frac{1}{a}i\hbar a' R^{a'} = -\frac{1}{a}i\hbar (-a) R^{-a} \\
& = i\hbar V_1^{-1}, \\
[V_2,(V_3+\tau V_1^{-1})] & = [V_2,V_3] + \tau[V_2,V_1^{-1}] + [V_2,\tau]V_1^{-1} \\
& = i\hbar V_3 + \tau i\hbar V_1^{-1} + 0 \\
& = i\hbar( V_3+\tau V_1^{-1})
\end{align}

obteremos as relações de comutação que definem a __álgebra de Lie $so(2,1)$__:
\begin{align}
[T_1,T_2] & = \Big[\frac{1}{2}\big(V_3 + \tau V_1^{-1} - V_1\big),V_2\Big] \\
& = \frac{1}{2}\big([V_3,V_2] + \tau[V_1^{-1},V_2] - [V_1,V_2]\big) \\
& = \frac{1}{2}\big(-i\hbar V_3 -\tau i\hbar V_1^{-1} - i\hbar V_1\big) \\
& = -i\hbar \frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big) \\
& = -i\hbar T_3, \\
[T_2,T_3] & = \Big[V_2,\frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big)\Big] \\
& = \frac{1}{2}\big( [V_2,V_3]  + \tau[V_2, V_1^{-1}] + [V_2,V_1]\big) \\
& = \frac{1}{2}\big( i\hbar V_3  + \tau i\hbar  V_1^{-1} -i\hbar V_1\big) \\
& = i\hbar\frac{1}{2}\big( V_3  + \tau  V_1^{-1} - V_1\big) \\
& = i\hbar T_1, \\
[T_3,T_1] & = \Big[\frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big), \frac{1}{2}\big(V_3 + \tau V_1^{-1} - V_1\big)\Big] \\
& = \frac{1}{4}\Big(0 + \tau[V_3,V_1^{-1}] - [V_3,V_1] + \tau[V_1^{-1},V_3] + 0 - \tau[V_1^{-1},V_1] + [V_1,V_3] + \tau[V_1,V_1^{-1}] + 0\Big) \\
& = \frac{1}{2}[V_1,V_3] \\
& = \frac{1}{2} 2i\hbar V_2 \\
& = i\hbar T_2,
\end{align}
em que usamos $[V_1,V_1^{-1}]=[R^a,R^{-a}]=0.$

Vemos assim que é possível escrever as relações de comutação que definem as álgebras $so(3)$ e $so(2,1)$ de forma unificada
\begin{align}
& [T_1,T_2] = \gamma i\hbar T_3, \\
& [T_2,T_3] = i\hbar T_1, \\
& [T_3,T_1] = i\hbar T_2,
\end{align}
com
\begin{align}
& \gamma = +1\ \rightarrow\ so(3), \\
& \gamma = -1\ \rightarrow\ so(2,1).
\end{align}

Seguindo, vamos definir o operador
$$T^2 := \gamma(T_1^2 + T_2^2) + T_3^2.$$
Para $\gamma=1$ esse operador faz o papel do operador para o quadrado do momento angular. Com essa definição, teremos que
\begin{align}
[T^2,T_1] & = [\gamma(T_1^2 + T_2^2) + T_3^2,T_1] \\
& = \gamma[T_1^2,T_1] + \gamma[T_2^2,T_1] + [T_3^2,T_1] \\
& = \gamma\times 0 + \gamma(T_2[T_2,T_1]+[T_2,T_1]T_2) + T_3[T_3,T_1] + [T_3,T_1]T_3 \\
& = \gamma(-T_2 \gamma i\hbar T_3-\gamma i\hbar T_3T_2) + T_3i\hbar T_2 + i\hbar T_2 T_3 \\
& = i\hbar(T_2 T_3 + T_3 T_2)(1-\gamma^2) \\
& = i\hbar(T_2 T_3 + T_3 T_2)(1-(\pm 1)^2) \\
& = i\hbar(T_2 T_3 + T_3 T_2)(1-1) \\
& = 0, \\
\end{align}

\begin{align}
[T^2,T_2] & = [\gamma(T_1^2 + T_2^2) + T_3^2,T_2] \\
& = \gamma[T_1^2,T_2] + \gamma[T_2^2,T_2] + [T_3^2,T_2] \\
& = \gamma(T_1[T_1,T_2]+[T_1,T_2]T_1) + 0 + T_3[T_3,T_2] + [T_3,T_2]T_3 \\
& = \gamma(T_1 \gamma i\hbar T_3+\gamma i\hbar T_3T_1) - T_3i\hbar T_1 - i\hbar T_1 T_3 \\
& = i\hbar(T_1 T_3 + T_3 T_1)(\gamma^2-1) \\
& = 0, \\
[T^2,T_3] & = [\gamma(T_1^2 + T_2^2) + T_3^2,T_3] \\
& = \gamma[T_1^2,T_3] + \gamma[T_2^2,T_3] + [T_3^2,T_3] \\
& = \gamma(T_1[T_1,T_3]+[T_1,T_3]T_1 + T_2[T_2,T_3] + [T_2,T_3]T_2) + 0 \\
& = \gamma(-T_1 i\hbar T_2 - i\hbar T_2T_1 + T_2 i\hbar T_1 + i\hbar T_1 T_2) \\
& = 0.
\end{align}

Vamos considerar os autovetores simultâneos de $T^2$ e $T_3$ sendo tais que seguem as seguintes equações de autovalores e autovetores:
\begin{align}
& T^2|Q,q\rangle = Q|Q,q\rangle, \\
& T_3|Q,q\rangle = q|Q,q\rangle.
\end{align}

Vamos definir os operadores de levantamento e abaixamento como:
$$T_\pm = T_1 \pm i T_2.$$
Teremos assim que
\begin{align}
[T^2,T_\pm] &= [T^2,T_1] \pm i [T^2,T_2] \\
& = 0.
\end{align}
Portanto
$$T^2(T_\pm|Q,q\rangle) = Q(T_\pm|Q,q\rangle).$$

Por outro lado
\begin{align}
[T_3,T_\pm] & = [T_3,T_1] \pm i [T_3,T_2] \\
& = i\hbar T_2 \pm i (-i\hbar T_1) \\
& = \pm\hbar(T_1 \pm i T_2) \\
& = \pm \hbar T_\pm \\
& = T_3 T_\pm - T_\pm T_3.
\end{align}

Teremos assim que
\begin{align}
T_3(T_\pm|Q,q\rangle) & = T_3T_\pm|Q,q\rangle \\
& = (\pm\hbar T_\pm + T_\pm T_3)|Q,q\rangle \\
& = \pm\hbar T_\pm|Q,q\rangle + T_\pm T_3|Q,q\rangle \\
& = \pm\hbar T_\pm|Q,q\rangle + T_\pm q|Q,q\rangle \\
& = (q\pm \hbar)|Q,q\rangle.
\end{align}
Ou seja, $T_\pm$ são os operadores escada para $T_3$, e por conseguinte podemos utilizar $T_\pm$ para obter os valores possíveis de $T_3$.

Notemos que
\begin{align}
\gamma T_\pm T_\mp & = \gamma(T_1 \pm i T_2)(T_1 \mp i T_2) \\
& = \gamma(T_1^2 \mp i T_1 T_2 \pm i T_2 T_1 + T_2^2) \\
& = \gamma(T_1^2 + T_2^2 \mp i [T_1, T_2]) \\
& = \gamma(T_1^2 + T_2^2 \mp i\times \gamma i\hbar T_3) \\
& = \gamma(T_1^2 + T_2^2) \pm \gamma^2\hbar T_3 \\
& = \gamma(T_1^2 + T_2^2) \pm \hbar T_3.
\end{align}
Portanto
\begin{align}
\frac{\gamma}{2}\big(T_+ T_- + T_- T_+\big) & = \gamma(T_1^2 + T_2^2) \\
& = \gamma(T_1^2 + T_2^2) + T_3^2 - T_3^2 \\
& = T^2 - T_3^2.
\end{align}

Vem assim que
\begin{align}
\langle Q,q|(T^2-T_3^2)|Q,q\rangle & = \langle Q,q|T^2|Q,q\rangle - \langle Q,q|T_3^2|Q,q\rangle \\
& = \langle Q,q|Q|Q,q\rangle - \langle Q,q|q^2|Q,q\rangle \\
& = Q - q^2 \\
& = \langle Q,q|\frac{\gamma}{2}\big(T_+ T_- + T_- T_+\big)|Q,q\rangle \\
& = \frac{\gamma}{2}\big(\langle Q,q|T_+ T_-|Q,q\rangle + \langle Q,q|T_- T_+|Q,q\rangle\big) \\
& =: \frac{\gamma}{2}\big(\langle \psi|\psi\rangle + \langle \phi|\phi\rangle\big),
\end{align}

em que definimos
\begin{align}
& |\psi\rangle := T_-|Q,q\rangle, \\
& |\phi\rangle := T_+|Q,q\rangle.
\end{align}

Como
\begin{align}
& \langle \psi|\psi\rangle \ge 0,\\
& \langle \phi|\phi\rangle \ge 0\\
& \therefore\ \langle \psi|\psi\rangle+\langle \phi|\phi\rangle \ge 0 \\
& \frac{2}{\gamma}(Q-q^2) = \langle \psi|\psi\rangle+\langle \phi|\phi\rangle \ge 0,
\end{align}
teremos os dois seguintes casos
\begin{align}
& \gamma = +1\ \therefore\ Q-q^2\ge 0\ \therefore\ q^2\le Q\ \therefore\ |q|\le\sqrt{Q}, \\
& \gamma = -1\ \therefore\ Q-q^2\le 0\ \therefore\ q^2\ge Q\ \therefore\ |q|\ge\sqrt{Q}.
\end{align}

Vemos assim que enquanto $so(3)$ é uma álgebra compacta (pois nesse caso $q$ é limitado superiormente e inferiormente: $q=-\sqrt{Q},-\sqrt{Q}+1,\cdots,\sqrt{Q}$), a álgebra $so(2,1)$ não é. Isso porque para $\gamma=-1$ temos que $q$ deve ter somente um limite superior ou um limite inferior, mas não ambos. Seguindo, escolhemos que $q$ tenha um limite inferior
$$q_\min := q_0. $$

Se atuarmos o operador de abaixamento no estado correspondente, como não podemos mais diminuir o valor de $T_3$, devemos obter o vetor nulo:
$$T_-|Q,q_0\rangle = |\oslash\rangle.$$
Usando essa equação e 
\begin{align}
\gamma T_\pm T_\mp & = \gamma(T_1^2 + T_2^2) \pm \hbar T_3 \\
& = \gamma(T_1^2 + T_2^2) + T_3^2 - T_3^2 \pm \hbar T_3 \\
& = T^2 - T_3^2 \pm \hbar T_3 \\
\end{align}
teremos que
\begin{align}
T^2|Q,q_0\rangle & = \big(\gamma T_+ T_- + T_3^2 - \hbar T_3\big)|Q,q_0\rangle \\
& = \gamma T_+ T_-|Q,q_0\rangle  + T_3^2|Q,q_0\rangle  - \hbar T_3|Q,q_0\rangle \\
& = \gamma T_+ |\oslash\rangle  + q_0^2|Q,q_0\rangle  - \hbar q_0|Q,q_0\rangle \\
& = q_0(q_0 - \hbar)|Q,q_0\rangle.
\end{align}
Ou seja, os valores possíveis de $T^2$ são 
$$Q=q_0(q_0-\hbar).$$

Para $\gamma=-1$, também podemos escrever
\begin{align}
& T^2 = -T_1^2 - T_2^2 + T_3^2 \\
& = - \frac{1}{4}\big(V_3 + \tau V_1^{-1} - V_1\big)\big(V_3 + \tau V_1^{-1} - V_1\big) - V_2^2 + \frac{1}{4}\big(V_3 + \tau V_1^{-1} + V_1\big)\big(V_3 + \tau V_1^{-1} + V_1\big) \\
& = - \frac{1}{4}\big(V_3^2 + \tau V_3 V_1^{-1} - V_3 V_1 + \tau V_1^{-1}V_3 + \tau^2 (V_1^{-1})^2 - \tau V_1^{-1}V_1 - V_1 V_3 - \tau V_1 V_1^{-1} + V_1^2\big) - V_2^2 \\
& \ + \frac{1}{4}\big(V_3^2 + \tau V_3 V_1^{-1} + V_3 V_1 + \tau V_1^{-1}V_3 + \tau^2 (V_1^{-1})^2 + \tau V_1^{-1}V_1 + V_1V_3 + \tau V_1 V_1^{-1} + V_1^2\big) \\
& = \frac{1}{2}\big( 0+ 0 + V_3 V_1+0+0 +\tau+ V_1V_3+\tau + 0\big) - V_2^2 \\
& = \tau + \frac{1}{2}(V_3 V_1+ V_1V_3) - V_2^2 \\
& = \tau + \frac{1}{2}(V_1 V_3 - 2i\hbar V_2  + V_1V_3) - V_2^2 \\
& = \tau + V_1 V_3 - i\hbar V_2 - V_2^2 \\
& = \tau + R^a a^{-2}R^{2-a}P^2 - i\hbar a^{-1}(RP - i\hbar(a-1)I/2) - a^{-2}(RP - i\hbar(a-1)I/2)(RP - i\hbar(a-1)I/2) \\
& = \tau + a^{-2}R^2P^2 - i\hbar a^{-1}RP -2^{-1}\hbar^2a^{-1}(a-1)I - a^{-2}RPRP + a^{-2}i\hbar(a-1)RP + a^{-2}2^{-2}\hbar^2(a-1)^2 I \\
& = \tau + a^{-2}R^2P^2 - i\hbar a^{-1}RP -2^{-1}\hbar^2(1-a^{-1})I - a^{-2}R(RP-i\hbar I)P + i\hbar(a^{-1} - a^{-2})RP + a^{-2}2^{-2}\hbar^2(a^2-2a+1) I \\
& = \tau + a^{-2}R^2P^2 - i\hbar a^{-1}RP -2^{-1}\hbar^2(1-a^{-1})I - a^{-2}R^2 P^2 + a^{-2}i\hbar RP + i\hbar(a^{-1} - a^{-2})RP + a^{-2}2^{-2}\hbar^2(a^2-2a+1) I \\
& = \tau  -2^{-1}\hbar^2(1-a^{-1})I + 2^{-2}\hbar^2(1-2a^{-1}+a^{-2}) I \\
& = \tau + \frac{\hbar^2}{4}(a^{-2}-1)I,
\end{align}

em que usamos $[V_1,V_3]=V_1 V_3 - V_3 V_1 = 2i\hbar V_2$, $V_1 = R^a$, $V_2 = a^{-1}(RP - i\hbar(a-1)I/2)$, $V_3=a^{-2}R^{2-a}P^2$, $T_1 = \frac{1}{2}\big(V_3 + \tau V_1^{-1} - V_1\big)$, $T_2 = V_2$, $T_3 = \frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big)$.

Com isso, teremos que
\begin{align}
\langle Q,q_0|T^2|Q,q_0\rangle & = \langle Q,q_0|Q|Q,q_0\rangle \\
& = q_0(q_0-\hbar) \\
& = \langle Q,q_0|\Big(\tau + \frac{\hbar^2}{4}(a^{-2}-1)I\Big)|Q,q_0\rangle \\
& = \tau + \frac{\hbar^2}{4}(a^{-2}-1).
\end{align}
Portanto
\begin{align}
& q_0^2 - \hbar q_0 -\tau - \frac{\hbar^2}{4}(a^{-2}-1) = 0 \\
& \therefore\ q_0 = \frac{1}{2}\Big(\hbar \pm \sqrt{\hbar^2 - 4\Big(-\tau - \frac{\hbar^2}{4}(a^{-2}-1)\Big)}\Big) \\ 
& \therefore\ q_0 = \frac{1}{2}\Big(\hbar \pm \sqrt{4\tau +\hbar^2 a^{-2}}\Big) \\ 
& \therefore\ q_0 = \frac{\hbar}{2}\Big(1 \pm \sqrt{\frac{4\tau}{\hbar^2} + \frac{1}{a^2}}\Big).
\end{align}

## Função de onda do estado fundamental

Usando as definições e resultados obtidos acima, teremos que
\begin{align}
& (T_- - T_3)|Q,q_0\rangle = T_-|Q,q_0\rangle - T_3|Q,q_0\rangle = |\oslash\rangle - q_0|Q,q_0\rangle =  - q_0|Q,q_0\rangle \\
& \therefore\ (-T_- + T_3 - q_0 I)|Q,q_0\rangle = (-T_1 + iT_2 + T_3 - q_0 I)|Q,q_0\rangle = |\oslash\rangle \\
& \therefore\ \Big(-\frac{1}{2}\big(V_3 + \tau V_1^{-1} - V_1\big) + iV_2 + \frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big) - q_0 I\Big)|Q,q_0\rangle = |\oslash\rangle \\
& \therefore\ \big(V_1 + iV_2 - q_0 I\big)|Q,q_0\rangle = |\oslash\rangle \\
& \therefore\ \big(R^a + i\frac{1}{a}\big(RP - i\hbar(a-1)I/2\big) - q_0 I\big)|Q,q_0\rangle = |\oslash\rangle \\
& \therefore\ \Big(R^a + \frac{i}{a}RP + \Big(\frac{\hbar(a-1)}{2a} - q_0\Big)I\Big)|Q,q_0\rangle = |\oslash\rangle \\
& \therefore\ \langle r|R^a|Q,q_0\rangle + \frac{i}{a}\langle r|RP|Q,q_0\rangle + \Big(\frac{\hbar(a-1)}{2a} - q_0\Big)\langle r|I|Q,q_0\rangle = \langle r|\oslash\rangle \\
& \therefore\ r^a\langle r|Q,q_0\rangle + \frac{i}{a}r\langle r|P|Q,q_0\rangle + \Big(\frac{\hbar(a-1)}{2a} - q_0\Big)\langle r|Q,q_0\rangle = 0 \\
& \therefore\ r^a \psi_0(r) + \frac{i}{a}r\frac{\hbar}{i}\frac{d}{dr}\langle r|Q,q_0\rangle + \Big(\frac{\hbar(a-1)}{2a} - q_0\Big)\psi_0(r) = 0 \\
& \therefore\ r^a \psi_0(r) + \frac{\hbar}{a}r\frac{d}{dr}\psi_0(r) + \Big(\frac{\hbar(a-1)}{2a} - q_0\Big)\psi_0(r) = 0 \\
& \therefore\ \Big( \frac{\hbar}{a}r\frac{d}{dr} + r^a  + \frac{\hbar(a-1)}{2a} - q_0\Big)\psi_0(r) = 0 \\
& \therefore\ \Big( r\frac{d}{dr} + \frac{a}{\hbar}r^a  + \frac{a-1}{2} - \frac{a}{\hbar}q_0\Big)\psi_0(r) = 0 \\
& \therefore\ \Big( r\frac{d\psi_0(r)}{dr} + \frac{a}{\hbar}r^a \psi_0(r)  + \Big(\frac{a-1}{2} - \frac{a}{\hbar}q_0\Big)\psi_0(r) = 0.
\end{align}

Dividindo essa última equação por $r\psi_0(r)/dr$ e integrando teremos
\begin{align}
& \int\frac{d\psi_0(r)}{\psi_0(r)} + \frac{a}{\hbar}\int r^{a-1}dr + \Big(\frac{a-1}{2} - \frac{a}{\hbar}q_0\Big)\int\frac{dr}{r} = 0 \\
& \therefore\ \ln(\psi_0(r)) - \ln(C) = \ln\Big(\frac{\psi_0(r)}{C}\Big) = -\frac{a}{\hbar}\frac{r^a}{a} - \Big(\frac{a-1}{2} - \frac{a}{\hbar}q_0\Big)\ln(r) \\
& \therefore\ e^{\ln\Big(\frac{\psi_0(r)}{C}\Big)} = e^{-\frac{r^a}{\hbar} - \Big(\frac{a-1}{2} - \frac{a}{\hbar}q_0\Big)\ln(r)} \\
& \therefore\ \psi_0(r) = Ce^{-r^a/\hbar}e^{-\Big(\frac{a-1}{2} - \frac{a}{\hbar}q_0\Big)\ln(r)} \\
& \therefore\ \psi_0(r) = Ce^{-r^a/\hbar}e^{\ln r^{\frac{1-a}{2} + \frac{aq_0}{\hbar}}} \\
& \therefore\ \psi_0(r) = Ce^{-r^a/\hbar}r^{(1-a)/2 + aq_0/\hbar}.
\end{align}

```python
from sympy import symbols, dsolve, Function, Derivative
r,hbb,q0 = symbols('r hbar q_0', positive=True)
a = symbols('a', positive=True)

psi0 = Function('psi0')
dsolve(r*Derivative(psi0(r),r) + ((a/hbb)*r**a)*psi0(r) + ((a-1)/2 - a*q0/hbb)*psi0(r), psi0(r))
```

## Funções de onda dos estados excitados

Para
$$q_n = q_0 + n_r \hbar,$$
temos que
$$T_3|Q,q_n\rangle = q_n|Q,q_n\rangle\ \therefore\ (T_3-q_n I)|Q,q_n\rangle = |\oslash\rangle.$$
Tendo
$$T_+|Q,q_n\rangle = k_{n+1}|Q,q_{n+1}\rangle$$
com
$$T_3|Q,q_{n+1}\rangle = q_{n+1}|Q,q_{n+1}\rangle,$$
teremos que
\begin{align}
& \big(T_+ - (T_3-q_n I)\big)|Q,q_n\rangle = T_+|Q,q_n\rangle = k_{n+1}|Q,q_{n+1}\rangle \\
& \therefore\ \big(T_1 + iT_2 - T_3 + q_n I\big)|Q,q_n\rangle = k_{n+1}|Q,q_{n+1}\rangle \\
& \therefore\ \Big(\frac{1}{2}\big(V_3 + \tau V_1^{-1} - V_1\big) + iV_2 - \frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big) + q_n I\Big)|Q,q_n\rangle = k_{n+1}|Q,q_{n+1}\rangle \\
& \therefore\ \Big(- V_1 + iV_2 + q_n I\Big)|Q,q_n\rangle = k_{n+1}|Q,q_{n+1}\rangle \\
& \therefore\ \big(-R^a + i\frac{1}{a}\big(RP - i\hbar(a-1)I/2\big) + q_n I\Big)|Q,q_n\rangle = k_{n+1}|Q,q_{n+1}\rangle \\
& \therefore\ -\langle r|R^a|Q,q_n\rangle + i\frac{1}{a}\langle r|RP|Q,q_n\rangle + \frac{1}{a}\hbar(a-1)\langle r|I|Q,q_n\rangle/2 + q_n \langle r|I|Q,q_n\rangle = k_{n+1}\langle r|Q,q_{n+1}\rangle \\
& \therefore\ -r^a\langle r|Q,q_n\rangle + \frac{i}{a}r\langle r|P|Q,q_n\rangle + \frac{1}{2a}\hbar(a-1)\langle r|Q,q_n\rangle + q_n \langle r|Q,q_n\rangle = k_{n+1}\psi_{n+1}(r) \\
& \therefore\ -r^a\psi_n(r) + \frac{i}{a}r\frac{\hbar}{i}\frac{d}{dr}\langle r|Q,q_n\rangle + \frac{1}{2a}\hbar(a-1)\psi_n(r) + q_n \psi_n(r) = k_{n+1}\psi_{n+1}(r) \\
& \therefore\ -r^a\psi_n(r) + r\frac{\hbar}{a}\frac{d}{dr}\psi_n(r) + \frac{1}{2a}\hbar(a-1)\psi_n(r) + q_n \psi_n(r) = k_{n+1}\psi_{n+1}(r) \\
& \therefore\ \psi_{n+1}(r) = \frac{1}{k_{n+1}}\Big(r\frac{d}{dr} - \frac{a r^a}{\hbar} + \frac{a-1}{2} + \frac{a q_n}{\hbar}\Big) \psi_n(r).
\end{align}

## Aplicação ao átomo de Hidrogênio


### Auto-energias

Consideremos a equação de autovalores e autovetores para o átomo de Hidrogênio
$$\Big(\frac{P^2}{2m} + \frac{\hbar^2 l(l+1)}{2mR^2} - \frac{ke^2}{R} - E\Big)|n,l,m_l\rangle = |\oslash\rangle.$$
Vamos multiplicar essa equação por $m\alpha R$ para obter
\begin{align}
& \Big(\frac{\alpha R P^2}{2} + \frac{\alpha\hbar^2 l(l+1)}{2R} - m\alpha ke^2 - m\alpha RE\Big)|n,l,m_l\rangle = |\oslash\rangle \\
& \therefore\ \Big(\frac{1}{2}\big(\alpha R P^2 + \alpha\hbar^2 l(l+1) R^{-1} - 2m\alpha ER\big) - m\alpha ke^2 \Big)|n,l,m_l\rangle = |\oslash\rangle.
\end{align}

Seguindo, vamos fazer as trocas de variáveis
\begin{align}
& \rho = \frac{R}{\alpha},\\
& \Xi=\alpha P
\end{align}
para obter
$$ \Big(\frac{1}{2}\big(\rho \Xi^2 + \hbar^2 l(l+1) \rho^{-1} - 2m\alpha E\rho\big) - m\alpha ke^2 \Big)|n,l,m_l\rangle = |\oslash\rangle.$$
Notemos que
$$[\rho,\Xi] = \Big[\frac{R}{\alpha},\alpha P\Big] = [R,P].$$

Se fizermos 
$$a=1$$
teremos
\begin{align}
& V_3 = \frac{1}{a^2}R^{2-a}P^2 = RP^2 \ \rightarrow\ V_3 = \rho\Xi^2,\\
& V_1 = R^a = R\ \rightarrow\ V_1 = \rho.
\end{align}
Com isso, obtemos
$$\Big(\frac{1}{2}\big(V_3 + \hbar^2 l(l+1) V_1^{-1} - 2m\alpha E V_1\big) - m\alpha ke^2 \Big)|n,l,m_l\rangle = |\oslash\rangle.$$
Essa equação é equivalente a 
$$(T_3-q)|n,l,m_l\rangle = |\oslash\rangle$$
se identificamos
$$T_3 = \frac{1}{2}\big(V_3 + \tau V_1^{-1} + V_1\big)$$
com
\begin{align}
& \tau = \hbar^2 l(l+1), \\
& - 2m\alpha^2 E = 1, \\
& m\alpha ke^2 = q.
\end{align}
Vale observar que a mudança de variáveis $\rho=R/\alpha,\ \Xi=\alpha P$ é necessária para conectar as 2 últimas equações acima.

Com isso, teremos 
\begin{align}
q_0 & = \frac{\hbar}{2}\Big(1 \pm \sqrt{\frac{4\tau}{\hbar^2} + \frac{1}{a^2}}\Big) \\
& = \frac{\hbar}{2}\Big(1 \pm \sqrt{\frac{4\hbar^2 l(l+1)}{\hbar^2} + \frac{1}{1^2}}\Big) \\
& = \frac{\hbar}{2}\Big(1 \pm \sqrt{4l(l+1)+1}\Big) \\
& = \frac{\hbar}{2}\Big(1 \pm (2l+1)\Big) \\
& := \frac{\hbar}{2}\Big(1 + (2l+1)\Big) \\
& = \hbar(l+1)
\end{align},
em que usamos o sinal positivo pois queremos que $q_0$ seja positivo. Com isso. vem que
\begin{align}
Q & = q_0(q_0-\hbar) \\
& = \hbar(l+1)\big(\hbar(l+1) - \hbar) \\
& = \hbar^2 l(l+1).
\end{align}

Sejam $n_r\hbar$ os incrementos de
\begin{align}
q & = q_0 + n_r\hbar = \hbar(l+1) + n_r\hbar \\
& = \hbar(n_r + l +1)
\end{align}
com 
$$n_r = 0,1,2,\cdots,$$
teremos que
\begin{align}
& m\alpha ke^2 = q = \hbar(n_r + l +1) \\
& \therefore \alpha = \frac{\hbar(n_r + l +1)}{mke^2}.
\end{align}
Com isso, obtemos uma expressão para a energia:
\begin{align}
& - 2m\alpha^2 E = 1 \\
& \therefore\ E = -\frac{1}{2m\alpha^2} = -\frac{mk^2 e^4}{2\hbar^2(n_r + l +1)^2} = -\frac{mk^2 e^4}{2\hbar^2 n^2},
\end{align}
que é a expressão de Bohr para a energia do átomo de Hidrogênio.
