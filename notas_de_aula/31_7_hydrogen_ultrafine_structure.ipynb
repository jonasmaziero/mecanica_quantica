# Estrutura ultrafina do Hidrogênio

Usando a relação relativística energia-momento, $E^2=P^2c^2+m^2c^4,$ podemos escrever a energia cinética como
$$E_{c} = E - E_0 = \frac{P^{2}}{2m} - \frac{P^{4}}{8m^{3}c^{2}} + \frac{P^{6}}{16m^{5}c^{4}} - \cdots.$$
A energia cinética Newtoniana, $P^{2}/2m,$ foi utilizada quando estudamos a estrutura principal de níveis de energia do Hidrogênio. O 2º termo, $- P^{4}/8m^{3}c^{2},$ foi utilizado para identificar a estrutura fina do Hidrogênio. Neste trabalho, você vai considerar o 3º termo, 
$$H_1^R = \frac{P^{6}}{16m^{5}c^{4}},$$
para obter a estrutura ultrafina do Hidrogênio. Para isso, aplique correção perturbativa de 1ª ordem:
$$E_n^{R(1)} = \langle\psi_{n,l,m_l}^{(0)}|H_1^R|\psi_{n,l,m_l}^{(0)}\rangle.$$
Verifique que essa correção pode ser escrita como
$$E_n^{R(1)} = \frac{n^6(E_n^{(0)})^3}{2 m^2 c^4}\Big(\frac{8}{(l+1/2)n^5} - \frac{8}{l(l+1/2)(l+1)n^3} - \frac{5}{n^6} + \frac{\langle UH_0U\rangle}{(E_n^{(0)})^3 n^6} \Big).$$
Ademais, verifique que o termo que não foi calculado explicitamente nessa expressão pode ser reescrito como
$$\langle UH_0U\rangle = \sum_{n',l',m_l'}k^2 e^4E_{n'}^{(0)}|\langle\psi_{n,l,m_l}| R^{-1}|\psi_{n',l',m_l'}\rangle|^2.$$

Verifique que o termo que determina a magnitude da correção pode ser reescrito e tem o seguinte valor aproximado:
$$\frac{n^6(E_n^{(0)})^3}{2m^2c^4}  = -\frac{\alpha^6 mc^2}{16} \approx -4.82\times 10^{-9}\ \text{eV}.$$
Como esse termo determina a magnitude dessa correção, notamos que ela determinará uma __estrutura ultrafina__ para o espectro de energias do átomo de Hidrogênio.

Começamos lembrando que o hamiltoniano não perturbado é
$$H_0 = \frac{P^2}{2m} + U\ \therefore\ P^2 = 2m(H_0 - U)$$
para escrever
\begin{align}
P^6 & = P^2 P^2 P^2 \\
& = 4m^2 P^2(H_0 -U)(H_0 - U) \\
& = 8m^3 (H_0 -U)(H_0^2 - H_0U - UH_0 + U^2) \\
& = 8m^3 (H_0^3 - H_0^2 U - H_0 UH_0 + H_0 U^2 - UH_0^2 + UH_0U + U^2 H_0 - U^3).
\end{align}
Com isso, usando $U = -ke^2/R$ e $H_0|\psi_{n,l,m_l}^{(0)}\rangle=E_n^{(0)}|\psi_{n,l,m_l}^{(0)}\rangle$, a correção perturbativa de 1ª ordem será dada por
\begin{align}
E_n^{R(1)} & = \langle\psi_{n,l,m_l}^{(0)}|H_1^R|\psi_{n,l,m_l}^{(0)}\rangle \\ 
& = \frac{1}{16 m^5 c^4}\langle\psi_{n,l,m_l}^{(0)}|P^6|\psi_{n,l,m_l}^{(0)}\rangle \\
& = \frac{8m^3}{16 m^5 c^4}\big(\langle\psi_{n,l,m_l}^{(0)}|H_0^3|\psi_{n,l,m_l}^{(0)}\rangle - \langle\psi_{n,l,m_l}^{(0)}|H_0^2 U|\psi_{n,l,m_l}^{(0)}\rangle - \langle\psi_{n,l,m_l}^{(0)}|H_0 UH_0|\psi_{n,l,m_l}^{(0)}\rangle + \langle\psi_{n,l,m_l}^{(0)}|H_0 U^2|\psi_{n,l,m_l}^{(0)}\rangle - \langle\psi_{n,l,m_l}^{(0)}|UH_0^2|\psi_{n,l,m_l}^{(0)}\rangle \\
& \hspace{2cm} + \langle\psi_{n,l,m_l}^{(0)}|UH_0U|\psi_{n,l,m_l}^{(0)}\rangle + \langle\psi_{n,l,m_l}^{(0)}|U^2 H_0|\psi_{n,l,m_l}^{(0)}\rangle - \langle\psi_{n,l,m_l}^{(0)}|U^3|\psi_{n,l,m_l}^{(0)}\rangle \big) \\
& = \frac{1}{2 m^2 c^4}\big((E_n^{(0)})^3 - (E_n^{(0)})^2\langle\psi_{n,l,m_l}^{(0)}|U|\psi_{n,l,m_l}^{(0)}\rangle - (E_n^{(0)})^2\langle\psi_{n,l,m_l}^{(0)}|U|\psi_{n,l,m_l}^{(0)}\rangle + E_n^{(0)}\langle\psi_{n,l,m_l}^{(0)}|U^2|\psi_{n,l,m_l}^{(0)}\rangle - (E_n^{(0)})^2\langle\psi_{n,l,m_l}^{(0)}|U|\psi_{n,l,m_l}^{(0)}\rangle \\
& \hspace{2cm} + \langle\psi_{n,l,m_l}^{(0)}|UH_0U|\psi_{n,l,m_l}^{(0)}\rangle + E_n^{(0)}\langle\psi_{n,l,m_l}^{(0)}|U^2|\psi_{n,l,m_l}^{(0)}\rangle - \langle\psi_{n,l,m_l}^{(0)}|U^3|\psi_{n,l,m_l}^{(0)}\rangle \big) \\
& = \frac{1}{2 m^2 c^4}\big((E_n^{(0)})^3 - 3(E_n^{(0)})^2\langle U\rangle  + 2E_n^{(0)}\langle U^2\rangle + \langle UH_0U\rangle - \langle U^3\rangle \big).
\end{align}

Vamos usar 
$$\langle R^{-1}\rangle=\frac{1}{an^{2}},\ \langle R^{-2}\rangle = \frac{1}{a^{2}n^{3}(l+1/2)},\ \langle R^{-3}\rangle = \frac{1}{l(l+1/2)(l+1)n^{3}a^{3}}$$
para obter
\begin{align}
\langle U\rangle & = -ke^2\langle R^{-1}\rangle = \frac{-ke^2}{an^{2}},\\
\langle U^2\rangle & = k^2 e^4\langle R^{-2}\rangle = \frac{k^2 e^4}{a^{2}n^{3}(l+1/2)},\\
\langle U^3\rangle & = -k^3 e^6\langle R^{-3}\rangle = \frac{-k^3 e^6}{l(l+1/2)(l+1)n^{3}a^{3}}.
\end{align}
O valor médio que nos falta resolver é
$$\langle UH_0U\rangle = k^2 e^4\langle R^{-1}H_0 R^{-1}\rangle.$$

Usando 
$$a := \frac{\hbar^{2}}{kme^{2}},\ E_{n}^{(0)} = -\frac{m k^{2}e^{4}}{2\hbar^{2}}\frac{1}{n^{2}} = -\frac{\alpha^{2}mc^{2}}{2n^{2}},$$
pode-se verificar que
\begin{align}
& - 3(E_n^{(0)})^2\langle U\rangle = -6(E_{n}^{(0)})^3,\\
& 2E_n^{(0)}\langle U^2\rangle = (E_{n}^{(0)})^3\frac{8n}{l+1/2},\\
& \langle U^3\rangle = (E_{n}^{(0)})^3\frac{8n^3}{l(l+1/2)(l+1)}.
\end{align}
Com isso, vem que
\begin{align}
E_n^{R(1)} & = \frac{1}{2 m^2 c^4}\Big((E_n^{(0)})^3 -6(E_{n}^{(0)})^3  + (E_{n}^{(0)})^3\frac{8n}{l+1/2} + \langle UH_0U\rangle - (E_{n}^{(0)})^3\frac{8n^3}{l(l+1/2)(l+1)} \Big) \\
& = \frac{n^6(E_n^{(0)})^3}{2 m^2 c^4}\Big(\frac{8}{(l+1/2)n^5} - \frac{8}{l(l+1/2)(l+1)n^3} - \frac{5}{n^6} + \frac{\langle UH_0U\rangle}{(E_n^{(0)})^3 n^6} \Big) \\
\end{align}

Ordem de grandeza da correção:
$$\frac{n^6(E_n^{(0)})^3}{2m^2c^4} = - \frac{n^6}{2m^2c^4}\frac{\alpha^6 m^3 c^6}{8 n^6} = -\frac{\alpha^6 mc^2}{16} \approx -4.82\times 10^{-9}\ \text{eV}.$$

As for the **ultrafine structure**, you're delving into even more subtle effects. While the fine and hyperfine structures account for corrections at the milli-electronvolt (meV) and micro-electronvolt (μeV) scales, the ultrafine structure involves even smaller energy differences.

The ultrafine structure arises from interactions beyond the previously mentioned effects. It includes contributions from:

1. **Lamb Shift**:
   - The Lamb shift, discovered by Willis Lamb and Robert Retherford in 1947, results from quantum electrodynamics (QED) effects.
   - It modifies the energy levels of hydrogen due to the interaction between the electron's magnetic moment and the vacuum fluctuations of the electromagnetic field.
   - The Lamb shift is responsible for a tiny correction to the energy levels, on the order of **10^-5 eV**[4].

2. **Vacuum Polarization**:
   - Vacuum polarization refers to the modification of the electromagnetic field around a charged particle due to the presence of virtual electron-positron pairs.
   - These virtual particles create an effective electric field that slightly alters the energy levels of the hydrogen atom.
   - The vacuum polarization effect contributes to the ultrafine structure, with corrections on the order of **10^-9 eV**[5].

3. **Self-Energy Corrections**:
   - Self-energy corrections account for the electron's interaction with its own electromagnetic field.
   - These corrections are also part of the ultrafine structure and contribute at the **10^-9 eV** level.

In summary, the ultrafine structure involves extremely subtle effects, including the Lamb shift, vacuum polarization, and self-energy corrections. While experimental measurements have confirmed these predictions, studying them requires sophisticated techniques and precision.

Researchers continue to explore these ultrafine effects, pushing the boundaries of our understanding of quantum electrodynamics and the fundamental nature of matter. The quest for ever more precise measurements and theoretical predictions remains an active area of research in quantum physics[5].

Certainly! The measurement of ultrafine effects in hydrogen involves sophisticated experimental techniques. While these effects are incredibly small, precise methods are essential for accurate characterization. Here are some techniques used:

1. **High-Resolution Spectroscopy**:
   - Ultrafine effects alter the energy levels of hydrogen atoms.
   - High-resolution spectroscopy allows scientists to study fine details in spectral lines.
   - Techniques like laser spectroscopy and Fourier-transform spectroscopy provide precise measurements of energy transitions.

2. **Cavity Ring-Down Spectroscopy (CRDS)**:
   - CRDS is an ultra-sensitive technique for measuring absorption or scattering of light.
   - It can detect tiny changes in energy levels due to ultrafine effects.
   - CRDS is used in precision measurements of atomic and molecular properties.

3. **Laser Cooling and Trapping**:
   - Laser cooling techniques allow researchers to cool hydrogen atoms to extremely low temperatures.
   - Cold atoms exhibit well-defined energy levels, making them ideal for studying ultrafine effects.
   - Techniques like magneto-optical trapping (MOT) and optical lattice trapping are used.

4. **Rydberg Atoms**:
   - Rydberg atoms have highly excited electronic states.
   - Their energy levels are sensitive to ultrafine corrections.
   - Researchers use Rydberg spectroscopy to explore fine details in hydrogen energy levels.

5. **Precision Atomic Clocks**:
   - Atomic clocks rely on stable energy levels.
   - Ultrafine effects impact clock accuracy.
   - Techniques like optical lattice clocks and trapped-ion clocks measure energy transitions with extreme precision.

6. **Quantum Monte Carlo Simulations**:
   - While not purely experimental, simulations play a role.
   - Quantum Monte Carlo methods provide accurate predictions of energy levels.
   - They help validate experimental results and guide further measurements.

7. **Muon Spin Spectroscopy**:
   - Muons are short-lived particles.
   - Muon spin spectroscopy studies their behavior in hydrogen-like systems.
   - It provides insights into ultrafine effects.

8. **Nuclear Magnetic Resonance (NMR)**:
   - NMR techniques probe nuclear energy levels.
   - Ultrafine corrections affect NMR spectra.
   - NMR is used in chemistry and materials science.

9. **Precision Mass Spectrometry**:
   - Mass spectrometers measure atomic and molecular masses.
   - Ultrafine effects influence mass differences.
   - High-precision mass spectrometry helps validate theoretical predictions.

10. **Synchrotron Radiation Experiments**:
    - Synchrotron facilities provide intense X-ray and ultraviolet radiation.
    - Researchers study hydrogen energy levels using synchrotron-based spectroscopy.
