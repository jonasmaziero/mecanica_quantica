# Entropia
Existe uma diferença fundamental entre a incerteza clássica associada à falta de informação sobre a preparação do sistema e a incerteza quântica associada com superposição. Esses dois tipos de incerteza podem coexistir dentro da descrição de estados via operador densidade. A seguir definiremos funções que podem ser utilizadas para quantificar essas incertezas.

## Entropia linear (EL)
Consideremos a decomposição espectral de um operador densidade, $\rho=\sum_{j=1}^{d}r_{j}|r_{j}\rangle\langle r_{j}|$. Vemos que o quadrado do operador densidade é
\begin{align}
\rho^{2} & = \sum_{j=1}^{d}r_{j}|r_{j}\rangle\langle r_{j}|\sum_{k=1}^{d}r_{k}|r_{k}\rangle\langle r_{k}| \\
& = \sum_{j,k=1}^{d}r_{j}r_{k}|r_{j}\rangle\langle r_{j}|r_{k}\rangle\langle r_{k}| \\
& = \sum_{j,k=1}^{d}r_{j}r_{k}|r_{j}\rangle\langle r_{k}|\delta_{j,k} \\
& = \sum_{j=1}^{d}r_{j}^{2}|r_{j}\rangle\langle r_{j}|.
\end{align}
Como $r_{j}\in[0,1]$ e nesse caso $r_{j}^{2}\le r_{j}$, temos que
\begin{align}
Tr(\rho^{2}) & = \sum_{j=1}^{d}r_{j}^{2}Tr(|r_{j}\rangle\langle r_{j}|) \\
& = \sum_{j=1}^{d}r_{j}^{2} \\
& \le \sum_{j=1}^{d}r_{j} \\
& = 1 \\
& = Tr(\rho).
\end{align}
No caso em que o estado do sistema é puro (não temos incerteza sobre a preparação), podemos considerar, sem perda de generalidade que 
$$\rho=|\psi_{1}\rangle\langle\psi_{1}|.$$
Assim $r_{1}^{2}=r_{1}=1$ e $r_{j\ne 1}=0$. Segue assim que nesse caso $Tr(\rho^{2})=Tr(\rho)=1$. No caso de um ensemble envolvendo estados de uma base ortonormal equiprováveis, $$\rho=\sum_{j=1}^{d}\frac{1}{d}|\phi_{j}\rangle\langle\phi_{j}|=\mathbb{I}_{d}/d,$$
temos máxima incerteza sobre a preparação e $Tr(\rho^{2}) = \sum_{j=1}^{d}(1/d)^{2} = d/d^{2} = 1/d < 1 = Tr(\rho)$. De forma geral temos
\begin{equation}
1/d \le Tr(\rho^{2}) \le 1.
\end{equation}

A __pureza__ de um operador densidade é definida com 
$$P(\rho) := \mathrm{Tr}(\rho^{2}).$$
Por sua vez, a EL é definida em termos da pureza como:
\begin{equation}
S_{L}(\rho) := 1-P(\rho)=1-Tr(\rho^{2}).
\end{equation}
Pode-se verificar que
\begin{equation}
0 = S_{L}(|\psi\rangle) \le S_{L}(\rho)\le S_{L}(\mathbb{I}/d) = (d-1)/d.
\end{equation}

## Entropia de Shannon
Para propor uma função que quantifique a informação que adquirimos ao tomar conhecimento do valor de uma variável aleatória, Shannon começou notando que a informação que obtemos ao tomar conhecimento do acontecimento de um evento $X_{j}$ é proporcional à nossa surpresa, que é inversamente proporcional a probabilidade daquele evento ocorrer:
\begin{equation}
\tilde{I}(X_{j}) = \frac{1}{p_{j}},
\end{equation}
com $p_{j}=Pr(X_{j})$. No entanto, para dois __eventos independentes__, $X_{j}$ e $Y_{k}$, para os quais $$Pr(X_{j},Y_{k})=p_{jk}=p_{j}p_{k}=Pr(X_{j})Pr(Y_{k}),$$
espera-se que a informação adquirida seja a soma das informações. Mas pela medida anterior
\begin{align}
\tilde{I}(X_{j},Y_{k}) & = \frac{1}{p_{jk}} = \frac{1}{p_{j}p_{k}} \\
& \ne \frac{1}{p_{j}} + \frac{1}{p_{k}} = I(X_{j})+I(Y_{k}).
\end{align}
O uso da função logaritmo nos fornece a esperada aditividade, i.e., definindo
\begin{equation}
I(X_{j})=\log_{2}\left(\frac{1}{p_{j}}\right) = \log_{2}\left(p_{j}^{-1}\right) = - \log_{2}(p_{j}).
\end{equation}
obtemos 
\begin{align}
I(X_{j},Y_{k}) & = -\log_{2}(p_{jk}) = -\log_{2}(p_{j}p_{k}) = -\log_{2}(p_{j})-\log_{2}(p_{k}) \\
& = I(X_{j})+I(Y_{k}).
\end{align}

Para uma variável aleatória $X$ que pode assumir os valores $X_{1},\cdots,X_{n}$ com probabilidades $p_{1},\cdots,p_{n}$, a quantidade de informação que obtemos ao tomar conhecimento do valor de $X$, ou, de forma equivalente, nossa incerteza sobre $X$ antes de saber do seu valor, pode ser obtida pela _média da informação_ que obtemos ao tomar conhecimento que cada um dos seus valores aconteceu:
\begin{equation}
H(X)=\sum_{j}p_{j}I(X_{j}) = -\sum_{j}p_{j}\log_{2}(p_{j}) \equiv H(\{p_j\}).
\end{equation}
Essa função é chamada de __entropia de Shannon__, e sua interpretação como o número mínimo de bits de informação necessário para armazenar a informação de uma fonte que emite símbolos $\{X_{j}\}$ com probabilidade $\{p_{j}\}$ é um dos esteios da teoria matemática da informação.

Vale observar que usamos $\log_{2}$ para medir entropia em bits.

Ademais, ao calcular $H$ usamos $\lim_{x\rightarrow 0} x\log x = 0$ para fazer $0\log 0 = 0$.

## Entropia de von Neumann (EvN)
Em suas investigações sobre entropia em sistemas termodinâmicos quânticos, von Neuman chegou na função
\begin{equation}
S_{vN}(\rho) = -\mathrm{Tr}(\rho\log_{2}\rho) = H(\{r_{j}\}),
\end{equation}
onde 
$$H(\{r_{j}\})=-\sum_{j}r_{j}\log_{2}r_{j}$$
é a __entropia de Shannon__ e $r_{j}$ são os autovalores de $\rho$. 

Benjamin Schumacher, em (https://doi.org/10.1103/PhysRevA.51.2738), forneceu uma interpretação operacional para a entropia de von Neuman $S_{vN}(\rho)$ como o número médio de sistemas quânticos de dois níveis (qbits) necessários para armazenar os estados da mistura em $\rho$.

Em termos de intervalos de valores, verifica-se que
\begin{equation}
0 = S_{vN}(|\psi\rangle) \le S_{vN}(\rho) \le S_{vN}(\mathbb{I}/d) = \log_{2}d.
\end{equation}

Vale mencionar que em termodinâmica usa-se $\ln=\log_{e}$ no lugar de $\log_{2}$ e a expressão da entropia é multiplicada por $k_{B}$, a constante de Boltzmann. Obtemos com isso a __entropia termodinâmica__:
\begin{equation}
S_{th}(\rho)=k_{B}\ln(2)S_{vN}(\rho)=-k_{B}Tr(\rho\ln\rho),
\end{equation}
que já consideramos anteriormente quando da discussão sobre estados térmicos.

### Um jogo para entender a EvN
Considere os observáveis $\sigma_{x}$ e $\sigma_{z}$ de um spin 1/2 preparado em um estado $\rho$, com medidas feitas usando aparatos de Stern-Gerlach. As __regras do jogo__ são as seguintes: Você deve escolher um único observável pra ser medido e um único autovetor desse observável. Cada vez que esse autovetor for obtido nas medidas você ganha um ponto, senão você perde um ponto. O objetivo do jogo é obter o maior saldo de pontos $N_{p}$ para um número total de medidas $N$. Vamos considerar alguns casos particulares:
* $\rho_{1}=|x_{+}\rangle\langle x_{+}|$: <br>
Nesse caso a melhor estratégia de jogo é escolher $\sigma_{x}$ e $|x_{+}\rangle$, pois nesse caso $S(|x_{+}\rangle)=0$ e sabemos que todas as medidas fornecerão o mesmo resultado, e assim $N_{p}=N_{x_{+}}-N_{x_{-}}=N-0$.

* $\rho_{2}=p_{1}|x_{+}\rangle\langle x_{+}|+p_{2}|x_{-}\rangle\langle x_{-}|$: <br>
Nesse caso, escolhendo-se $\sigma_{x}$, a melhor estratégia depende do valor de $p_{1}$. Se $p_{1}\ge 1/2$ escolhemos $|x_{+}\rangle$ obtendo 
$$N_{p} = N_{x_+}-N_{x_-} = p_{1}N-p_{2}N=(2p_{1}-1)N.$$ 
Se $p_{1}< 1/2$ escolhe-se $|x_{-}\rangle$ obtendo $N_{p}=(2p_{2}-1)N$ pontos. A EvN desse estado é 
$$S_{vN}(\rho_{2})=H(\{p_{1},1-p_{2}\})=-p_{1}\log_{2}(p_{1})-(1-p_{1})\log_{2}(1-p_{1}).$$

* $\rho_{3}=p_{1}|x_{+}\rangle\langle x_{+}|+p_{2}|z_{+}\rangle\langle z_{+}|$: <br>
Se $p_{1}\ge 1/2$ escolhemos $\sigma_{x}$ e $|x_{+}\rangle$, e assim $N_{p}=p_{1}N+p_{2}(1/2)N-p_{2}(1/2)N=p_{1}N$. Se $p_{1}<1/2$, escolhemos $\sigma_{z}$ e $|z_{+}\rangle$, obtendo $N_{p}=p_{1}(1/2)N-p_{1}(1/2)N+p_{2}N=p_{2}N$. Nesse caso a EvN é 
$$S_{vN}(\rho_{3})=-\lambda_{1}\log_{2}(\lambda_{1})-(1-\lambda_{1})\log_{2}(1-\lambda_{1})$$
com $\lambda_{1}=(1+\sqrt{1-2p_{1}(1-p_{1})})/2$.

Veja nos gráficos abaixo a relação "inversa" que existe entre o número de pontos que fazemos e a entropia de von Neumann do estado quântico.

```python
import numpy as np
from matplotlib import pyplot as plt

def jogo():
    p1 = np.arange(10**-6,1,0.001); 
    Np1 = p1-p1+1; S1 = 0*p1
    Np2 = np.maximum(2*p1-1,2*(1-p1)-1); 
    S2 = -p1*np.log2(p1)-(1-p1)*np.log2(1-p1)
    Np3 = np.maximum(p1,1-p1)
    l1 = (1+np.sqrt(1-2*p1*(1-p1)))/2; 
    S3 = -l1*np.log2(l1)-(1-l1)*np.log2(1-l1)
    plt.plot(p1,Np1,'--',color='blue'); 
    plt.plot(p1,S1,color='blue')
    plt.plot(p1,Np2,'--',color='red'); 
    plt.plot(p1,S2,color='red')
    plt.plot(p1,Np3,'--',color='black'); 
    plt.plot(p1,S3,color='black')
    plt.legend((r'$N_{p}(\rho_{1})$',r'$S_{vN}(\rho_{1})$',
                r'$N_{p}(\rho_{2})$',r'$S_{vN}(\rho_{2})$',
               r'$N_{p}(\rho_{3})$',r'$S_{vN}(\rho_{3})$'),
               loc=(1.01,0.2), fontsize=13)
    plt.xlabel(r'$p_{1}$'); plt.xlim(0,1); 
    plt.ylim(-0.02,1.02)
    plt.show()
jogo()
```

## Entropia de Wigner-Yanase (EWY)
Enquanto as EL e EvN quantificam incerteza clássica, a EWY quantifica incerteza quântica em relação a um observável $O$ para um sistema preparado no estado $\rho$:
\begin{align}
S_{wy}(O|\rho) & := -\frac{1}{2}Tr\left([\sqrt{\rho},O]^{2}\right) \\
& = -\frac{1}{2}Tr\left((\sqrt{\rho}O-O\sqrt{\rho})^{2}\right) \\
& = -\frac{1}{2}Tr\left( (\sqrt{\rho}O-O\sqrt{\rho})(\sqrt{\rho}O-O\sqrt{\rho}) \right) \\
& = -\frac{1}{2}Tr\left( \sqrt{\rho}O\sqrt{\rho}O -O\sqrt{\rho}\sqrt{\rho}O - \sqrt{\rho}OO\sqrt{\rho} +O\sqrt{\rho}O\sqrt{\rho} \right) \\
& = -\frac{1}{2}\left( Tr(\sqrt{\rho}O\sqrt{\rho}O) - Tr(O\rho O) - Tr(\sqrt{\rho}O^{2}\sqrt{\rho}) + Tr(O\sqrt{\rho}O\sqrt{\rho}) \right) \\
& = Tr(\rho O^{2}) - Tr(\sqrt{\rho}O\sqrt{\rho}O) \\
& = \langle O^{2}\rangle_{\rho} - Tr\left((O\sqrt{\rho})^{2}\right).
\end{align}

Note que se $\rho$ e $O$ comutam, a EWY é nula, mesmo que nossa incerteza clássica sobre o estado do sistema seja positiva.

Se o estado for puro, i.e., se $\rho=|\psi\rangle\langle\psi|$, pode-se verificar que a entropia de Wigner-Yanase coincide com a variância 
$$E_{wy}(|\psi\rangle)=Var_{\psi}(O).$$

### Exemplo
Comecemos verificando que o comprimento do vetor de Bloch não muda por transformações unitárias. Para verificar essa informação, considera $\rho'=U\rho U^{\dagger}$. Para sistemas de 2 níveis
\begin{align}
P(\rho) & = Tr \left(\frac{1}{2}\left(\sigma_{0} + \sum_{j=1}^{3}r_{j}\sigma_{j}\right)\frac{1}{2}\left(\sigma_{0} + \sum_{k=1}^{3}r_{k}\sigma_{k}\right)\right) \\
& = \frac{1}{4}Tr\left(\sigma_{0} + 2\sum_{j=1}^{3}r_{j}\sigma_{j} + \sum_{j,k=1}^{3}r_{j}r_{k}\sigma_{j}\sigma_{k} \right) \\
& = \frac{1}{4}\left(2 + 0 + \sum_{j,k=1}^{3}r_{j}r_{k}Tr(\sigma_{j}\sigma_{k}) \right) \\
& = \frac{1}{4}\left(2 + \sum_{j,k=1}^{3}r_{j}r_{k}2\delta_{jk} \right) \\
& = \frac{1}{2}\left(1 + r^{2} \right).
\end{align}
Pela invariância da função traço sob transformações unitárias (mudança de base), teremos que
\begin{align}
P(\rho') & = \frac{1}{2}\left(1 + r'^{2} \right) \\
& = Tr(U\rho U^{\dagger}U\rho U^{\dagger}) \\
& = Tr(U\rho\rho U^{\dagger})  \\
& = Tr(\rho^{2}) \\
& = P(\rho).
\end{align}
Portanto $r'=r$. Ou seja, tranformações unitárias somente giram o vetor de Bloch em $\mathbb{R}^{3}$.

Vamos calcular aquelas três entropias para o estado geral de um sistema de 2 níveis mostrado acima. Da expressão para a pureza vem que
\begin{equation}
S_{L}(\rho) = \frac{1-r^{2}}{2}.
\end{equation}
Já, dos autovalores de $\rho$, a EN fica
\begin{align}
S_{vN}(\rho) & = -\frac{1+r}{2}\log_{2}\frac{1+r}{2} - \frac{1-r}{2}\log_{2}\frac{1-r}{2} \\
             & = 1 - 2^{-1}[(1+r)\log_{2}(1+r) + (1-r)\log_{2}(1-r)].
\end{align}


Para obter $S_{wy}$ consideraremos como observável $\sigma_{z}$. Nesse caso, como $\sigma_{z}^{2}=\sigma_{0}$, fazendo $\rho=r_{+}|r_{+}\rangle\langle r_{+}|+r_{-}|r_{-}\rangle\langle r_{-}|$ com $r_{\pm}=(1\pm r)/2$ e $|r_{+}\rangle = \cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle$ e $|r_{-}\rangle = \sin(\theta/2)|z_{+}\rangle - e^{i\phi}\cos(\theta/2)|z_{-}\rangle$. Assim, usando $\sigma_{z}|r_{+}\rangle = \cos(\theta/2)|z_{+}\rangle - e^{i\phi}\sin(\theta/2)|z_{-}\rangle$,  $\sigma_{z}|r_{-}\rangle=\sin(\theta/2)|z_{+}\rangle + e^{i\phi}\cos(\theta/2)|z_{-}\rangle$ e $Tr(A|\psi\rangle\langle\psi|)=\langle\psi|A|\psi\rangle$ obteremos
\begin{align}
S_{wy}(\sigma_{z}|\rho) & = Tr(\rho) - Tr\left((\sigma_{z}\sqrt{\rho})^{2}\right) \\
& = 1 - Tr\left((\sigma_{z}(\sqrt{r_{+}}|r_{+}\rangle\langle r_{+}|+\sqrt{r_{-}}|r_{-}\rangle\langle r_{-}|))^{2}\right) \\
& = 1 - Tr(\sqrt{r_{+}}\sigma_{z}|r_{+}\rangle\langle r_{+}| + \sqrt{r_{-}}\sigma_{z}|r_{-}\rangle\langle r_{-}|)^{2} \\
& = 1 - Tr\left(r_{+}\sigma_{z}|r_{+}\rangle\langle r_{+}|\sigma_{z}|r_{+}\rangle\langle r_{+}| + \sqrt{r_{+}r_{-}}\sigma_{z}|r_{+}\rangle\langle r_{+}|\sigma_{z}|r_{-}\rangle\langle r_{-}| + \sqrt{r_{+}r_{-}}\sigma_{z}|r_{-}\rangle\langle r_{-}|\sigma_{z}|r_{+}\rangle\langle r_{+}| + r_{-}\sigma_{z}|r_{-}\rangle\langle r_{-}|\sigma_{z}|r_{-}\rangle\langle r_{-}| \right) \\
& = 1 - r_{+}\langle r_{+}|\sigma_{z}|r_{+}\rangle^{2} - 2\sqrt{r_{+}r_{-}}|\langle r_{-}|\sigma_{z}|r_{+}\rangle|^{2} - r_{-}\langle r_{-}|\sigma_{z}|r_{-}\rangle^{2} \\
& = 1 - r_{+}(\cos^{2}(\theta/2)-\sin^{2}(\theta/2))^{2} - 2\sqrt{r_{+}r_{-}}|2\sin(\theta/2)\cos(\theta/2)|^{2} - r_{-}(\sin^{2}(\theta/2)-\cos^{2}(\theta/2))^{2} \\
& = 1 - 2^{-1}(1+r)\cos^{2}\theta - \sqrt{1-r^{2}}\sin^{2}\theta - 2^{-1}(1-r)\cos^{2}\theta  = 1 - \cos^{2}\theta - \sqrt{1-r^{2}}\sin^{2}\theta \\
& = (1-\sqrt{1-r^{2}})\sin^{2}\theta.
\end{align}

```python
from ipywidgets import interactive
import math

def entropies(n):
    th = n*np.pi/8; N = 100; dr = (1.01-0)/N
    rv = np.zeros(N); P = np.zeros(N); Sl = np.zeros(N)
    Sn = np.zeros(N); 
    Swy = np.zeros(N); #th = entropies(th)
    r = -dr
    for j in range(0,N):
        r += dr; rv[j] = r; P[j] = (1+r**2)/2; Sl[j] = 1-P[j]
        Sn[j] = 1-(1/2)*((1+r)*math.log(1+r,2) + (1-r)*math.log(1-r,2)); 
        Swy[j] = (1-math.sqrt(1-r**2))*(math.sin(th))**2
    plt.xlabel(r'$r$'); plt.plot(rv, P, label=r'$P$'); 
    plt.plot(rv, Sl, label=r'$S_{L}$')
    plt.plot(rv, Sn, label=r'$S_{vN}$'); 
    plt.plot(rv, Swy, label=r'$S_{WY}(\theta=\pi/4)$')
    axes = plt.gca(); axes.set_xlim([0,1]); 
    axes.set_ylim([-0.1,1.1])
    plt.legend(bbox_to_anchor=(1.05, 1), 
               loc='upper left', borderaxespad=0.); 
    plt.show()
interactive(entropies, n=(0,8))
```
