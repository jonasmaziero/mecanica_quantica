# Átomos Hidrogenóides

Vamos começar considerando o problema de um núcleo ($Z$ prótons) interagindo com um elétron via somente a interação coulombiana. Mesmo assim esse problema, que envolve forças centrais, é de 2 corpos. Para resolver esse problema, primeiramente transformaremos esse problema de 2 corpos em um problema equivalente de uma partícula efetiva. Claro, nesse caso o potencial não depende do tempo e queremos resolver a equação de Schrödinger independente do tempo na base da posição:
$$H\psi(\vec{r}) = E\psi(\vec{r}),$$
com 
$$H = K + U$$
para 
$$U = U(r).$$

Se for necessário, podemos obter a função de onda dependente do tempo:
$$\Psi(\vec{r},t) = \psi(\vec{r})e^{-iEt/\hbar}.$$

## Forças centrais: $2\rightarrow 1$ corpo
Aqui estaremos interessados no caso de um elétron interagindo via interação Coulombiana com um núcleo com $Z$ prótons, que corresponde a uma energia potencial elétrica
\begin{equation}
U(r)=-\frac{kZe^{2}}{r},
\end{equation}
onde $\vec{r}$ é a posição do elétron em relação ao núcleo e 
$$k=\frac{1}{4\pi\epsilon_{0}}$$
é a constante de Coulomb, com $\epsilon_{0}$ sendo a permissividade do vácuo. 

Vale observar, no entanto, que os resultados desta seção se aplicam para qualquer potencial central $U = U(r).$
A energia total do sistema, clássico, é então
\begin{equation}
H= \frac{m_{e}v_{e}^{2}}{2} + \frac{m_{n}v_{n}^{2}}{2} + U(r),
\end{equation}
em que 
\begin{align}
& \vec{v}_{e}=\partial_{t}\vec{r}_{e}, \\
& \vec{v}_{n}=\partial_{t}\vec{r}_{n}
\end{align}
com $\vec{r}_{e}$ e $\vec{r}_{n}$ sendo, respectivamente, as posições do elétron e do núcleo em relação a um referencial inercial genérico $O$ e $\partial_t = \frac{\partial}{\partial_t}$. Veja a figura abaixo.

<img src='fig_16_referencial.png' width='500'>

A posição do centro de massa desse sistema em relação a $O$ é 
$$\vec{r}_{cm} = \frac{m_{e}}{M}\vec{r}_{e}+\frac{m_{n}}{M}\vec{r}_{n},$$
com a massa total sendo
$$M=m_{e}+m_{n}.$$
Como nesse caso 
$$\sum\vec{F}_{externas}=\vec{0},$$
o centro de massa pode ser usado como um referencial inercial. Fazendo 
$$O=CM$$
teremos
\begin{equation}
\vec{r}_{cm}=\vec{0} \ \therefore\  m_{e}\vec{r}_{e}=-m_{n}\vec{r}_{n}.
\end{equation}
Além disso, também temos que
\begin{equation}
\vec{r} = -\vec{r}_{n}+\vec{r}_{e}.
\end{equation}

Substituindo uma dessas duas últimas equações na outra, e derivando em relação ao tempo, obteremos
\begin{align}
& \vec{r} = -\vec{r}_{n}-\frac{m_{n}}{m_{e}}\vec{r}_{n} = -\frac{m_{e}+m_{n}}{m_{e}}\vec{r}_{n} \ \therefore\ -\vec{v}_{n} = \frac{m_{e}}{m_{e}+m_{n}}\vec{v}, \\
& \vec{r} =  \frac{m_{e}}{m_{n}}\vec{r}_{e}+\vec{r}_{e} =  \frac{m_{e}+m_{n}}{m_{n}}\vec{r}_{e} \ \therefore\  \vec{v}_{e} = \frac{m_{n}}{m_{e}+m_{n}}\vec{v}.
\end{align}
Substituindo esses resultados na equação para a energia, vem que
\begin{align}
H &= \frac{1}{2}m_{e}\frac{m_{n}^{2}}{(m_{e}+m_{n})^{2}}v^{2} + \frac{1}{2}m_{n}\frac{m_{e}^{2}}{(m_{e}+m_{n})^{2}}v^{2} + U(r) \\
& = \frac{1}{2}\frac{m_{e}m_{n}}{(m_{e}+m_{n})^{2}}v^{2}(m_{n}+m_{e}) + U(r) \\
& =: \frac{1}{2}\mu v^{2} + U(r) \\
& \equiv \frac{p^{2}}{2\mu}+U(r),
\end{align}
em que definimos a __massa reduzida__ do sistema elétron-núcleo como
\begin{equation}
\mu = \frac{m_{e}m_{n}}{m_{e}+m_{n}}.
\end{equation}
e denotamos o momento linear por 
$$\vec{p}=\mu\partial_{t}\vec{r}.$$
Com isso, transformamos o problema de dois corpos (envolvendo dois vetores posição $\vec{r}_{e}$ e $\vec{r}_{n}$) em um problema de um único corpo (envolvendo um único vetor posição $\vec{r}$).

Como o hamiltoniano, depois de quantizado, depende de $P^2 = -\hbar^2\nabla^2$ e o problema que estamos considerando aqui tem simetria esférica, a seguir vamos obter o operador laplaciano em coordenadas esféricas. Para fazer isso, vamos primeiramente aprender como obter operadores diferencias em um sistema de coordenadas curvilíneas qualquer.

## Laplaciano em coordenadas esféricas

Quantizando o momento linear, faremos a troca
$$p\rightarrow -i\hbar\vec{\nabla}$$
na relação clássica para o hamiltoniano. Assim, para a energia de átomos de um elétron, obteremos a seguinte equação de Schrödinger independente (EqSIT) do tempo tridimensional:
\begin{equation}
-\frac{\hbar^{2}}{2\mu}\nabla^{2}\psi(x,y,z)+U(r)\psi(x,y,z)=E\psi(x,y,z),
\end{equation}
com $\vec{r}=(x,y,z)$ e o Laplaciano é dado por
\begin{equation}
\nabla^{2} = \partial_{xx}+\partial_{yy}+\partial_{zz}
\end{equation}
e 
$$r=\sqrt{x^{2}+y^{2}+z^{2}}.$$
Acima usamos a notação $\partial_{xx}=\frac{\partial^2}{\partial x^2},\ \partial_{yy}=\frac{\partial^2}{\partial y^2},\ \partial_{zz}=\frac{\partial^2}{\partial z^2}.$ As coordenadas $x,y,z$ são aquelas do elétron em relação ao núcleo. No entanto, como a energia potencial possui simetria esférica, é vantajoso escrever a EqSIT em coordenadas esféricas, o que faremos na sequência.

### Operador gradiente em coordenadas curvilíneas
O sistema de coordenadas cartesianas é definido pelas superfícies 
$$x_{1}=x=\text{constante},\ x_{2}=y=\text{constante},\ x_{3}=z=\text{constante}.$$
Cada trio de coordenadas corresponde à intersecção entre esses três planos. Nesse caso, os vetores que fornecem as direções ortogonais a esses planos, $$\hat{x}_{1}=\hat{i},\ \hat{x}_{2}=\hat{j},\ \hat{x}_{3}=\hat{k},$$
são ortogonais e de módulo constante para qualquer ponto do espaço tri-dimensional. 

Para problemas tais como o que estamos considerando aqui, é útil utilizar outros sistemas de coordenadas. Consideremos um sistema de __coordenadas curvilíneas__, definido pelas superfícies
\begin{equation}
\{q_{j}=\text{constante}\}_{j=1}^{3}.
\end{equation}
Os vetores $\hat{q}_{j}$ ortogonais a essas superfícies não precisam ter direção constante. As coordenadas dos dois sistemas podem, a princípio, ser relacionadas:
\begin{equation}
q_{j} = q_{j}(\{x_{k}\}) \text{ e } x_{k} = x_{k}(\{q_{j}\})\text{, com } j,k=1,2,3.
\end{equation}
Assim, podemos escrever a diferencial
\begin{equation}
dx_{k} = \sum_{j=1}^{3}\frac{\partial x_{k}}{\partial q_{j}}dq_{j}\text{, para } k=1,2,3,
\end{equation}
da qual obtemos o __deslocamento infinitesimal quadrático__ ($\vec{r}$ é o vetor posição em $\mathbb{R}^{3}$):
\begin{align}
ds^{2} & = d\vec{r}\cdot d\vec{r} \\
& = \sum_{k=1}^{3}\hat{x}_{k}dx_{k}\cdot\sum_{l=1}^{3}\hat{x}_{l}dx_{l} \\
& = \sum_{k,l=1}^{3}(\hat{x}_{k}\cdot \hat{x}_{l})dx_{k}dx_{l} \\
& = \sum_{k,l=1}^{3}\delta_{k,l}dx_{k}dx_{l} \\
& = \sum_{k=1}^{3}dx_{k}dx_{k} \\
& = \sum_{k=1}^{3}\sum_{j=1}^{3}\frac{\partial x_{k}}{\partial q_{j}}dq_{j}\sum_{l=1}^{3}\frac{\partial x_{k}}{\partial q_{l}}dq_{l} \\
& = \sum_{j,l=1}^{3}\left(\sum_{k=1}^{3}\frac{\partial x_{k}}{\partial q_{j}}\frac{\partial x_{k}}{\partial q_{l}}\right)dq_{j}dq_{l} \\
& = \sum_{j,l=1}^{3}\frac{\partial \vec{r}}{\partial q_{j}}\cdot\frac{\partial \vec{r}}{\partial q_{l}}dq_{j}dq_{l} \\
& =: \sum_{j,l=1}^{3}g_{jl}dq_{j}dq_{l},
\end{align}

em que definimos o __tensor métrico de Riemann__ como
\begin{equation}
g_{jl}:=\frac{\partial \vec{r}}{\partial q_{j}}\cdot\frac{\partial \vec{r}}{\partial q_{l}}.
\end{equation}

Notando que uma __mudança infinitesimal de uma coordenada $q_{j}$__ corresponde a um deslocamento infinitesimal (considere e.g. as coordenadas polares em 2D, para as quais se $\theta$ muda de $d\theta$, o deslocamento infinitesimal é igual a $rd\theta$):
\begin{equation}
ds_{j}:=h_{j}dq_{j},
\end{equation}
em que $h_{j}$ pode depender de $q_{j}$ e das outras coordenadas $q_{k\ne j}$, escrevemos
\begin{equation}
d\vec{r}=\sum_{j=1}^{3}\hat{q}_{j}ds_{j}=\sum_{j=1}^{3}\hat{q}_{j}h_{j}dq_{j},
\end{equation}
e assim teremos
\begin{equation}
ds^{2} =  d\vec{r}\cdot d\vec{r} = \sum_{j,l=1}^{3} (h_{j}\hat{q}_{j})\cdot(h_{l}\hat{q}_{l})dq_{j}dq_{l}  = \sum_{j,l=1}^{3} h_{j}h_{l}(\hat{q}_{j}\cdot\hat{q}_{l})dq_{j}dq_{l} = \sum_{j,l=1}^{3} g_{jl}dq_{j}dq_{l}.
\end{equation}
Por conseguinte
\begin{equation}
h_{j}\hat{q}_{j} = \frac{\partial \vec{r}}{\partial q_{j}}.
\end{equation}

Assim, se considerarmos um __sistema de coordenadas ortogonais__, i.e., 
\begin{equation}
\hat{q}_{j}\cdot\hat{q}_{l}=0 \text{ para } j\ne l,
\end{equation}
teremos também 
$$g_{jl}= h_j h_l (\hat{q}_{j}\cdot\hat{q}_{l}) = 0 \ \text{para}\ j\ne l$$ 
e
\begin{equation}
ds^{2} = \sum_{j=1}^{3}g_{jj}dq_{j}^{2} = \sum_{j=1}^{3}h_{j}^{2}dq_{j}^{2},
\end{equation}
ou
\begin{equation}
h_{j} = \sqrt{g_{jj}} = \sqrt{\sum_{k=1}^{3}\left(\frac{\partial x_{k}}{\partial q_{j}}\right)^{2}}.
\end{equation}

Assim, o vetor __gradiente__ da função escalar $\psi(\{q_{j}\})$ fica escrito como
\begin{equation}
\vec{\nabla}\psi(\{q_{j}\})=\sum_{j=1}^{3}\hat{q}_{j}\frac{\partial\psi}{\partial s_{j}}=\sum_{j=1}^{3}\hat{q}_{j}\frac{1}{h_{j}}\frac{\partial\psi}{\partial q_{j}},
\end{equation}
e fornece a amplitude e direção de máxima variação da função $\psi$ no ponto $\{q_{j}\}$.

#### Operador gradiente em coordenadas esféricas
Em coordenadas esféricas temos 
$$q_{1}=r,\ q_{2}=\theta,q_{3}=\phi$$
e
\begin{align}
& x=x_{1}=r\sin\theta\cos\phi, \\
& y=x_{2}=r\sin\theta\sin\phi, \\
& z=x_{3}=r\cos\theta.
\end{align}
Por conseguinte
\begin{align}
h_{1}^{2} & = g_{11} \\
& = \left(\frac{\partial x_{1}}{\partial r}\right)^{2}+\left(\frac{\partial x_{2}}{\partial r}\right)^{2}+\left(\frac{\partial x_{3}}{\partial r}\right)^{2} \\ 
&= \left(\sin\theta\cos\phi\right)^{2}+\left(\sin\theta\sin\phi\right)^{2}+\left(\cos\theta\right)^{2} \\
& = 1, \\
h_{2}^{2} & = g_{22} \\
& = \left(\frac{\partial x_{1}}{\partial \theta}\right)^{2}+\left(\frac{\partial x_{2}}{\partial \theta}\right)^{2}+\left(\frac{\partial x_{3}}{\partial \theta}\right)^{2} \\ 
&= \left(r\cos\theta\cos\phi\right)^{2}+\left(r\cos\theta\sin\phi\right)^{2}+\left(-r\sin\theta\right)^{2} \\
& = r^{2}, \\
h_{3}^{2} & = g_{33} \\
& = \left(\frac{\partial x_{1}}{\partial \phi}\right)^{2}+\left(\frac{\partial x_{2}}{\partial \phi}\right)^{2}+\left(\frac{\partial x_{3}}{\partial \phi}\right)^{2} \\ 
&= \left(-r\sin\theta\sin\phi\right)^{2}+\left(r\sin\theta\cos\phi\right)^{2}+\left(0\right)^{2} \\
& = r^{2}\sin^{2}\theta.
\end{align}
O operador gradiente em coordenadas esféricas é dado como segue:
\begin{align}
\vec{\nabla} & = \hat{q}_{1}\frac{1}{h_{1}}\frac{\partial}{\partial q_{1}} + \hat{q}_{2}\frac{1}{h_{2}}\frac{\partial}{\partial q_{2}} + \hat{q}_{3}\frac{1}{h_{3}}\frac{\partial}{\partial q_{3}} \\
& = \hat{r}\frac{1}{1}\frac{\partial}{\partial r} + \hat{\theta}\frac{1}{r}\frac{\partial}{\partial \theta} + \hat{\phi}\frac{1}{r\sin\theta}\frac{\partial}{\partial \phi} \\
& = \hat{r}\frac{\partial}{\partial r} + \hat{\theta}\frac{1}{r}\frac{\partial}{\partial \theta} + \hat{\phi}\frac{1}{r\sin\theta}\frac{\partial}{\partial \phi}.
\end{align}

### Operador Laplaciano em coordenadas curvilíneas
Seguindo, vamos fazer uso do teorema de Gauss
\begin{equation}
\oint_{S_{V}}\vec{v}\cdot d\vec{A} = \int_{V}\vec{\nabla}\cdot\vec{v}dV.
\end{equation}
Quando aplicado a um elemento infinitesimal de volume, 
$$dV=ds_{1}ds_{2}ds_{3}=h_{1}dq_{1}h_{2}dq_{2}h_{3}dq_{3},$$
esse teorema implica que:
\begin{equation}
\sum_{faces}\vec{v}\cdot d\vec{A} = \vec{\nabla}\cdot\vec{v}dV,
\end{equation}
em que $\vec{v}$ é um campo vetorial. Focando em uma direção particular, se mudamos a coordenada $q_{1}$ de uma quantidade infinitesimal $dq_{1}$, teremos (veja a figura abaixo):
\begin{align}
(v_{1}dA_{1})_{\text{direita}} + (v_{1}dA_{1})_{\text{esquerda}} &=\frac{\left(v_{1}h_{2}dq_{2}h_{3}dq_{3}\right)_{q_{1}+dq_{1}}-\left(v_{1}h_{2}dq_{2}h_{3}dq_{3}\right)_{q_{1}}}{dq_{1}}dq_{1} \\
&=\frac{\left(v_{1}h_{2}h_{3}\right)_{q_{1}+dq_{1}}-\left(v_{1}h_{2}h_{3}\right)_{q_{1}}}{dq_{1}}dq_{1}dq_{2}dq_{3} \\
& = \frac{\partial(v_{1}h_{2}h_{3})}{\partial q_{1}}dq_{1}dq_{2}dq_{3},
\end{align}
em que usamos
$$dA_1 = h_2 dq_2 h_3 dq_3.$$

<img src='fig_16_gauss.png' width='400'>

Do mesmo modo, pode-se mostrar que
\begin{align}
& \sum v_2 dA_2 = \frac{\partial(v_{2}h_{1}h_{3})}{\partial q_{2}}dq_{1}dq_{2}dq_{3}, \\
& \sum v_3 dA_3 = \frac{\partial(v_{3}h_{1}h_{2})}{\partial q_{3}}dq_{1}dq_{2}dq_{3}.
\end{align}

Somando os fluxos nas três direções, obteremos o divergente de um campo vetorial qualquer $\vec{v}$ em um sistema de coordenadas curvilíneas qualquer $\{q_j\}$:
\begin{align}
\vec{\nabla}\cdot\vec{v} & = \frac{\sum_{faces}\vec{v}\cdot d\vec{A}}{dV} \\
& = \frac{\sum v_{1}dA_{1}+\sum v_{2}dA_{2}+\sum v_{3}dA_{3}}{dV} \\
& = \frac{ \frac{\partial(v_{1}h_{2}h_{3})}{\partial q_{1}} + \frac{\partial(v_{2}h_{1}h_{3})}{\partial q_{2}} + \frac{\partial(v_{3}h_{1}h_{2})}{\partial q_{3}} }{h_{1}h_{2}h_{3}dq_{1}dq_{2}dq_{3}}dq_{1}dq_{2}dq_{3} \\
& = \frac{1}{h_{1}h_{2}h_{3}}\left(\frac{\partial(v_{1}h_{2}h_{3})}{\partial q_{1}} + \frac{\partial(v_{2}h_{1}h_{3})}{\partial q_{2}} + \frac{\partial(v_{3}h_{1}h_{2})}{\partial q_{3}}\right).
\end{align}

Usando como campo vetorial o gradiente
$$\vec{v}=\vec{\nabla}\psi = \sum_j \hat{q}_j \frac{1}{h_j}\frac{\partial\psi}{\partial q_j},$$
obteremos o Laplaciano
\begin{align}
\nabla^{2}\psi & = \vec{\nabla}\cdot\vec{\nabla}\psi \\
& = \frac{1}{h_{1}h_{2}h_{3}}\left(\frac{\partial}{\partial q_{1}}\left(\frac{h_{2}h_{3}}{h_{1}}\frac{\partial\psi}{\partial q_{1}}\right) + \frac{\partial}{\partial q_{2}}\left(\frac{h_{1}h_{3}}{h_{2}}\frac{\partial\psi}{\partial q_{2}}\right) + \frac{\partial}{\partial q_{3}}\left(\frac{h_{1}h_{2}}{h_{3}}\frac{\partial\psi}{\partial q_{3}}\right) \right).
\end{align}

#### Operador Laplaciano em coordenadas esféricas
Em coordenadas esféricas 
$$h_{1}=1,\ h_{2}=r,\ h_{3}=r\sin\theta$$
e o operador Laplaciano fica escrito como
\begin{align}
\nabla^{2} & = \frac{1}{r^{2}\sin\theta}\left(\frac{\partial}{\partial r}\left(r^{2}\sin\theta\frac{\partial}{\partial r}\right) + \frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial}{\partial \theta}\right) + \frac{\partial}{\partial \phi}\left(\frac{1}{\sin\theta}\frac{\partial}{\partial \phi}\right) \right) \\
& = \frac{1}{r^{2}}\frac{\partial}{\partial r}\left(r^{2}\frac{\partial}{\partial r}\right) + \frac{1}{r^{2}\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial}{\partial \theta}\right) + \frac{1}{r^{2}\sin^{2}\theta}\frac{\partial^{2}}{\partial \phi^{2}}.
\end{align}

## EqSIT em coordenadas esféricas
A EqSIT em coordenadas esféricas fica escrita como ($\psi=\psi(r,\theta,\phi)$)
\begin{align}
& \frac{-\hbar^{2}}{2\mu}\nabla^{2}\psi(r,\theta,\phi)+U(r)\psi(r,\theta,\phi)=E\psi(r,\theta,\phi) \\
& \therefore\ \frac{-\hbar^{2}}{2\mu}\left(\frac{1}{r^{2}}\frac{\partial}{\partial r}\left(r^{2}\frac{\partial\psi}{\partial r}\right) + \frac{1}{r^{2}\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial\psi}{\partial \theta}\right) + \frac{1}{r^{2}\sin^{2}\theta}\frac{\partial^{2}\psi}{\partial \phi^{2}} \right) + U(r)\psi = E\psi.
\end{align}

Tendo em vista a forma da energia potencial, $U=U(r)$, vamos usar __separação de variáveis__,
\begin{equation}
\psi(r,\theta,\phi) = \Omega(r)\Upsilon(\theta,\phi),
\end{equation}
para resolver a EqSIT, que fica escrita como
\begin{align}
& \frac{-\hbar^{2}}{2\mu}\left(\Upsilon(\theta,\phi)\frac{1}{r^{2}}\frac{\partial}{\partial r}\left(r^{2}\frac{\partial\Omega(r)}{\partial r}\right) + \frac{\Omega(r)}{r^{2}\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial\Upsilon(\theta,\phi)}{\partial \theta}\right) + \frac{\Omega(r)}{r^{2}\sin^{2}\theta}\frac{\partial^{2}\Upsilon(\theta,\phi)}{\partial \phi^{2}} \right) \\
& + U(r)\Omega(r)\Upsilon(\theta,\phi) = E\Omega(r)\Upsilon(\theta,\phi).
\end{align}
Manipulando essa última equação, obteremos
\begin{align}
& \frac{1}{\Omega(r)}\frac{\partial}{\partial r}\left(r^{2}\frac{\partial\Omega(r)}{\partial r}\right) + \frac{r^{2}2\mu(E-U(r))}{\hbar^{2}} \\
& = -\frac{1}{\Upsilon(\theta,\phi)\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial\Upsilon(\theta,\phi)}{\partial \theta}\right) - \frac{1}{\Upsilon(\theta,\phi)\sin^{2}\theta}\frac{\partial^{2}\Upsilon(\theta,\phi)}{\partial \phi^{2}} \\ 
& = l(l+1),
\end{align}
onde a última igualdade foi incluída numa constante $l(l+1)$, que vem do fato das variáveis $r$ e $(\theta,\phi)$ serem independentes.

A equação a ser resolvida para obtermos a parte __radial__ da função de onda é
\begin{equation}
\frac{\partial}{\partial r}\left(r^{2}\frac{\partial\Omega(r)}{\partial r}\right) + \left(\frac{r^{2}2\mu(E-U(r))}{\hbar^{2}}-l(l+1)\right)\Omega(r)=0.
\end{equation}

Seguindo, aplicamos separação de variáveis para a parte angular, 
$$\Upsilon(\theta,\phi)=\Theta(\theta)\Phi(\phi),$$
da equação acima:
\begin{align}
& \frac{1}{\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial\Theta(\theta)\Phi(\phi)}{\partial \theta}\right) + \frac{1}{\sin^{2}\theta}\frac{\partial^{2}\Theta(\theta)\Phi(\phi)}{\partial \phi^{2}} + l(l+1)\Theta(\theta)\Phi(\phi) = 0, \\
& \therefore \frac{\sin\theta}{\Theta(\theta)}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial\Theta(\theta)}{\partial \theta}\right) + l(l+1)\sin^{2}\theta = -\frac{1}{\Phi(\phi)}\frac{\partial^{2}\Phi(\phi)}{\partial \phi^{2}} \\
& \hspace{0.5cm} = m_{l}^{2},
\end{align}
em que introduzimos a constante $m_{l}^{2}$ pois $\theta$ e $\phi$ são variáveis independentes. 

A equação referente à função de onda do ângulo __longitudinal__ é
\begin{equation}
\frac{\partial^{2}\Phi(\phi)}{\partial \phi^{2}}+m_{l}^{2}\Phi(\phi)=0.
\end{equation}

A equação diferencial para a função de onda do ângulo __latitudinal__ fica escrita como
\begin{equation}
\sin\theta\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial\Theta(\theta)}{\partial \theta}\right) + \left(l(l+1)\sin^{2}\theta-m_{l}^{2}\right)\Theta(\theta)=0.
\end{equation}
