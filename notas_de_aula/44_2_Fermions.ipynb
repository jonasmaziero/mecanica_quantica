# Relações de comutação canônicas para férmions

Considere operadores lineares $a_1,\cdots,a_n$ atuando em um espaço de Hilbert $V$ de dimensão finita:
$$a_1,\cdots,a_n:V\rightarrow V.$$
Dizemos que esses operadores satisfazem as relações de comutação canônicas para férmions (RCCF) se
\begin{align}
& \{a_j,a_k^\dagger\} = \delta_{j,k}I \\
& \{a_j,a_k\}=0_V,
\end{align}
em que 
$$j,k=1,\cdots,n$$
e
$$\{A,B\} = AB+BA = \{B,A\}$$
é o anti-comutador. Acima, $I$ é o operador identidade em $V$ e $0_V$ é o operador nulo em $V$.

Notemos que 
$$\{a_j,a_k\}=0\ \therefore\ \{a_j,a_k\}^\dagger=0_V^\dagger\ \therefore\ \{a_j^\dagger,a_k^\dagger\}=0_V.$$
Além disso, considerando $j=k$ teremos que
\begin{align}
& \{a_j,a_j\}=0_V\ \therefore\ a_j^2 =0_V,\\
& \{a_j^\dagger,a_j^\dagger\}=0_V\ \therefore\ (a_j^\dagger)^2=0_V.
\end{align}

#### Exemplo:
Considere o operador de abaixamento para um qubit (um spin 1/2):
$$a = |0\rangle\langle 1|\ \therefore\ a^\dagger=|1\rangle\langle 0|.$$
É fácil verificar que
\begin{align}
& \{a,a^\dagger\} = aa^\dagger + a^\dagger a = |0\rangle\langle 1||1\rangle\langle 0| + |1\rangle\langle 0||0\rangle\langle 1| = |0\rangle\langle 0| + |1\rangle\langle 1| = I, \\
& \{a,a\} = 2a^2 = 2|0\rangle\langle 1||0\rangle\langle 1| = 0_V, \\
& a^2 = |0\rangle\langle 1||0\rangle\langle 1| = 0_V, \\
& (a^\dagger)^2 = |1\rangle\langle 0||1\rangle\langle 0 = 0_V.
\end{align}

## Consequências das RCCF

### Proposição
Os operadores $$N_j := a_j^\dagger a_j,$$ para $$j=1,\cdots,n,$$ são positivos semi-definidos e tem autovalores $0$ e $1$.

#### Prova
Um operador linear $A:V\rightarrow V$ é dito positivo semi-definido se $$\langle A\rangle_\psi = \langle\psi|A|\psi\rangle\ge 0\ \forall\ |\psi\rangle\in V.$$ Para $N_j$ teremos que 
\begin{align}
\langle N_j\rangle_\psi & = \langle\psi|N_j|\psi\rangle = \langle\psi|a_j^\dagger a_j|\psi\rangle \\
& = \big(|\psi\rangle,a_j^\dagger a_j|\psi\rangle\big) = \big(a_j|\psi\rangle, a_j|\psi\rangle\big) \\
& = \big(|\xi_j\rangle, |\xi_j\rangle\big) \\
& \ge 0,
\end{align}
em que definimos 
$$|\xi_j\rangle=a_j|\psi\rangle$$
e usamos a positividade do produto interno, $\langle\xi_j|\xi_j\rangle\ge 0\ \forall\ |\xi_j\rangle\in V$, a existência do adjunto, $(|\psi\rangle,A|\phi\rangle) = (A^\dagger|\psi\rangle,|\phi\rangle)\ \forall\ A:V\rightarrow V,\ |\psi\rangle,|\phi\rangle\in V.$ Agora, usando $\{a_j,a_j^\dagger\}=a_j a_j^\dagger + a_j^\dagger a_j = I$, $a_j^2=(a_j^\dagger)^2=0_V$ e a equação de autovalores e autovetores $$N_j|n_j\rangle = a_j^\dagger a_j|n_j\rangle = n_j|n_j\rangle$$ teremos que
\begin{align}
N_j^2|n_j\rangle & = n_j^2 |n_j\rangle \\
& = a_j^\dagger a_j a_j^\dagger a_j|n_j\rangle \\
& = a_j^\dagger (I- a_j^\dagger a_j) a_j|n_j\rangle \\
& = a_j^\dagger a_j|n_j\rangle - a_j^\dagger a_j^\dagger a_j a_j|n_j\rangle \\
& = N_j|n_j\rangle - (0_V)(0_V)|n_j\rangle \\
& = N_j|n_j\rangle - |\oslash\rangle \\
& = n_j|n_j\rangle,
\end{align}
em que $|\oslash\rangle$ é o vetor nulo. Por conseguinte 
\begin{align}
& n_j^2 = n_j \\
& \therefore\ n_j=0\ \text{ou}\ n_j=1.
\end{align}

$\square$

### Proposição
O operador $a_j$ ($a_j^\dagger$) atua como um operador de abaixamento (levantamento) para $N_j$.

#### Prova
Vamos considerar primeiramente um autovetor de $N_j$ correspondente ao autovalor $1$: $$N_j|n_j(1)\rangle = (1)|n_j(1)\rangle.$$ Teremos assim que 
\begin{align}
N_j(a_j|n_j(1)\rangle) & = a_j^\dagger a_j a_j|n_j(1)\rangle \\
& = (I- a_j a_j^\dagger)a_j|n_j(1)\rangle \\
& = a_j|n_j(1)\rangle - a_j N_j|n_j(1)\rangle \\
& = a_j|n_j(1)\rangle - a_j (1)|n_j(1)\rangle \\
& = (1-1)(a_j|n_j(1)\rangle) \\
& = (0)(a_j|n_j(1)\rangle).
\end{align}
Note que acima poderíamos ter usado $a_j^2 = 0_V$, mas isso não nos traria nenhuma informação relevante. 

Do resultado acima teremos a seguinte ação explícita do operador de abaixamento:
\begin{align}
& a_j|n_j(1)\rangle \propto |n_j(0)\rangle =: c|n_j(0)\rangle \\
& \therefore\ \langle n_j(1)| a_j^\dagger a_j|n_j(1)\rangle = \langle n_j(1)| N_j|n_j(1)\rangle = \langle n_j(1)|(1)|n_j(1)\rangle = 1 \\
& = |c|^2\langle n_j(0)|n_j(0)\rangle = |c|^2 \\
& \therefore c=1 \\
& \therefore\ a_j|n_j(1)\rangle = |n_j(0)\rangle.
\end{align}
Usando $(a_j^\dagger)^2 =0_V$ também teremos que
\begin{align}
& N_j(a_j^\dagger|n_j(1)\rangle) = a_j^\dagger a_j a_j^\dagger|n_j(1)\rangle = a_j^\dagger(I- a_j^\dagger a_j)|n_j(1)\rangle \\
& = a_j^\dagger|n_j(1)\rangle - a_j^\dagger N_j|n_j(1)\rangle = a_j^\dagger|n_j(1)\rangle - a_j^\dagger (1)|n_j(1)\rangle = (1-1)a_j^\dagger|n_j(1)\rangle \\
& = (0)a_j^\dagger|n_j(1)\rangle = |\oslash\rangle \\
& = a_j^\dagger|n_j(1)\rangle - a_j^\dagger a_j^\dagger a_j|n_j(1)\rangle = a_j^\dagger|n_j(1)\rangle - 0_V a_j|n_j(1)\rangle = a_j^\dagger|n_j(1)\rangle - |\oslash\rangle \\
& = a_j^\dagger|n_j(1)\rangle.
\end{align}
Portanto $$a_j^\dagger|n_j(1)\rangle = |\oslash\rangle.$$
Já para um autovetor de $N_j$ com autovalor $0$, $$N_j|n_j(0)\rangle = (0)|n_j(0)\rangle,$$ usando $a_j^2 = 0_V$, teremos
\begin{align}
N_j(a_j|n_j(0)\rangle & = a_j^\dagger a_j a_j|n_j(0)\rangle = a_j^\dagger 0_V|n_j(0)\rangle = a_j^\dagger |\oslash\rangle \\
& = |\oslash\rangle \\
& = (I- a_j a_j^\dagger)a_j|n_j(0)\rangle = a_j|n_j(0)\rangle - a_j N_j|n_j(0)\rangle = a_j|n_j(0)\rangle - a_j (0)|n_j(0)\rangle = a_j|n_j(0)\rangle - a_j |\oslash\rangle \\
& = a_j|n_j(0)\rangle.
\end{align}
Portanto $$a_j|n_j(0)\rangle = |\oslash\rangle.$$
No caso de $a_j^\dagger$ teremos
\begin{align}
N_j(a_j^\dagger|n_j(0)\rangle) & = a_j^\dagger a_j a_j^\dagger|n_j(0)\rangle \\
& = a_j^\dagger(I-  a_j^\dagger a_j)|n_j(0)\rangle \\
& = a_j^\dagger|n_j(0)\rangle - a_j^\dagger N_j|n_j(0)\rangle \\
& = a_j^\dagger|n_j(0)\rangle - a_j^\dagger (0)|n_j(0)\rangle \\
& = (1)(a_j^\dagger|n_j(0)\rangle) + a_j^\dagger|\oslash\rangle \\
& = (1)(a_j^\dagger|n_j(0)\rangle).
\end{align}
Assim a ação explícita do operador de levantamento fica
\begin{align}
& a_j^\dagger|n_j(0)\rangle \propto |n_j(1)\rangle =: c|n_j(1)\rangle \\
& \therefore\ \langle n_j(0)|a_j a_j^\dagger|n_j(0)\rangle = \langle n_j(0)|(I- a_j^\dagger a_j)|n_j(0)\rangle \\
& =  \langle n_j(0)|n_j(0)\rangle -  \langle n_j(0)|N_j|n_j(0)\rangle = 1 - \langle n_j(0)|(0)|n_j(0)\rangle \\
& = |c|^2\langle n_j(1)|n_j(1)\rangle = |c|^2 \\
& \therefore\ c =1 \\
& \therefore\ a_j^\dagger|n_j(0)\rangle = |n_j(1)\rangle.
\end{align}

$\square$

### Proposição

Os operadores $\{N_j\}_{j=1}^{n} = \{a_j^\dagger a_j\}_{j=1}^{n}$ formam um conjunto mutuamente compatível, e portanto compartilham uma mesma base de autovetores.

#### Prova 

Vamos usar as RCCF $$\{a_j,a_k^\dagger\} = a_j a_k^\dagger + a_k^\dagger a_j = \delta_{j,k}I$$ e $$\{a_j,a_k\}=a_j a_k + a_k a_j = 0$$ e $$\{a_j^\dagger,a_k^\dagger\}=a_j^\dagger a_k^\dagger + a_k^\dagger a_j^\dagger = 0$$ para obter
\begin{align}
[N_j, N_k] & = [a_j^\dagger a_j,a_k^\dagger a_k] \\
& = a_j^\dagger a_j a_k^\dagger a_k - a_k^\dagger a_k a_j^\dagger a_j \\
& = a_j^\dagger (\delta_{j,k}I - a_k^\dagger a_j ) a_k - a_k^\dagger (\delta_{k,j}I - a_j^\dagger a_k ) a_j \\
& = a_j^\dagger a_k \delta_{j,k} - a_j^\dagger a_k^\dagger a_j a_k - a_k^\dagger a_j \delta_{k,j}  + a_k^\dagger a_j^\dagger a_k a_j \\
& = a_j^\dagger a_j - (-)a_k^\dagger a_j^\dagger a_j a_k - a_j^\dagger a_j  + (-)a_k^\dagger a_j^\dagger a_j a_k \\
& = 0_V.
\end{align}
Por conseguinte, existe uma base ortonormal de $V$ que diagonaliza todos os operadores $\{N_j\}_{j=1}^n$.

$\square$

### Proposição

Existe um conjunto de $2^n$ estados ortonormais que são autovetores comuns aos $\{N_j\}_{j=1}^n$ e que podem ser obtidos a partir do estado $$|n_1(0)n_2(0)\cdots n_n(0)\rangle := |0 0 \cdots 0\rangle_{12\cdots n}$$ aplicando os operadores de levantamento.

#### Prova

Cada modo fermiônico $|j\rangle_s$, para 
$$s=1,\cdots,n,$$
pode assumir 2 estados, 
$$|0\rangle_s \text{ ou } |1\rangle_s,$$
analogamente a um qubit. Existe um férmion no modo $s$ ou nenhum. Assim como para $n$ qubits, temos $2^n$ possíveis estados da base computacional desses modos:
\begin{align}
& |00\cdots 00\rangle_{12\cdots(n-1)n}, \\
& |00\cdots 01\rangle_{12\cdots(n-1)n}, \\
& |00\cdots 10\rangle_{12\cdots(n-1)n}, \\
& \vdots \\
& |01\cdots 00\rangle_{12\cdots(n-1)n}, \\
& |10\cdots 00\rangle_{12\cdots(n-1)n}\\
& \vdots \\
& |01\cdots 11\rangle_{12\cdots(n-1)n}, \\
& |11\cdots 11\rangle_{12\cdots(n-1)n}.
\end{align}
Considere $n_j$ o autovalor de $N_j$ e defina o estado de vácuo (nenhum modo ocupado) como 
\begin{align}
|vac\rangle=|00\cdots 0\rangle_{12\cdots n}.
\end{align}
Qualquer estado do conjunto de $2^n$ estados mostrado acima pode ser obtido usando os operadores de levantamento ou a identidade como segue: 
\begin{align}
|n\rangle &= |n_1 n_2 \cdots n_n\rangle_{12\cdots n} \\
& = (a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots (a_n^\dagger)^{n_n}|vac\rangle_{12\cdots n}.
\end{align}
Esse conjunto de vetores gera um espaço vetorial $W$, que é um subespaço de $V$. E já vimos que 
\begin{align}
& a_j : W\rightarrow W, \\
& a_j^\dagger : W\rightarrow W.
\end{align}

$\square$

### Proposição
Seja $W_\perp$ um complemento ortogonal de $W$:
$$W_\perp = \{|v\rangle\in V\ |\ \langle v|w\rangle=0\ \forall\ |w\rangle\in W\}.$$ Então
\begin{align}
& a_j : W_\perp\rightarrow W_\perp, \\
& a_j^\dagger : W_\perp\rightarrow W_\perp.
\end{align}

#### Prova

Consideremos $|\psi\rangle\in W_\perp$ e $|\phi\rangle\in W$. Teremos assim que
\begin{align}
\langle\psi|a_j|\phi\rangle & = (|\psi\rangle,a_j|\phi\rangle) = 0 \\
& = (a_j^\dagger|\psi\rangle,|\phi\rangle) = (|\phi\rangle,a_j^\dagger|\psi\rangle)^* =0.
\end{align}
A primeira igualdade a zero segue pois $a_j:W\rightarrow W$. Assim $a_j|\phi\rangle\in W$ é ortogonal a $|\psi\rangle.$ Também usamos a existência do adjunto para ir para a 2ª linha. A última igualdade só pode ser obtida de forma geral se $a_j^\dagger|\psi\rangle$ for ortogonal a $|\phi\rangle$, o que implica que 
$$a_j^\dagger:W_\perp\rightarrow W_\perp.$$

Também temos que
\begin{align}
\langle\psi|a_j^\dagger|\phi\rangle & = (|\psi\rangle,a_j^\dagger|\phi\rangle) = 0 \\
& = ((a_j^\dagger)^\dagger|\psi\rangle,|\phi\rangle) = (a_j|\psi\rangle,|\phi\rangle) = (|\phi\rangle,a_j|\psi\rangle)^* =0.
\end{align}
A primeira igualdade a zero segue pois $a_j \daggerˆ:W\rightarrow W$. Assim $a_j^\dagger|\phi\rangle\in W$ é ortogonal a $|\psi\rangle.$ Também usamos a existência do adjunto para ir para a 2ª linha. A última igualdade só pode ser obtida de forma geral se $a_j|\psi\rangle$ for ortogonal a $|\phi\rangle$, o que implica que 
$$a_j:W_\perp\rightarrow W_\perp.$$
$\square$

OBS. A prova acima não está totalmente satisfatória pra mim, a menos que $W\oplus W_\perp = V.$

Sendo a prova válida, as RCCF também seriam válidas em $W_\perp.$

Iterando esse processo de construção de subespaços ortogonais, como $V$ tem dimensão finita, depois de digamos $d$ iterações teremos completado o espaço vetorial como um todo:
$$W_1 \oplus W_2 \oplus \cdots \oplus W_d = V,$$
lembrando que
$$\dim(W_j)=2^n.$$
Com isso, podemos introduzir uma base
$$|\alpha,k\rangle\in V$$
com 
$$\alpha=1,\cdots,2^n,\ k=1,\cdots,d.$$ 
Dessa forma, podemos considerar
$$V=\mathbb{C}^{2^n}\otimes \mathbb{C}^d$$
com a ação trivial de $a_j,\ a_j^\dagger$ em $\mathbb{C}^d.$$
Essa é a chamada __representação do número de ocupação__ ou __álgebra de Fermi__.

Em termos da linguagem de teoria de grupos, quando 
$$d=1$$
dizemos ter obtido um representação fundamental das RCCF.

### Proposição

A ação dos operadores de abaixamento em $W$, mais especificamente no estado $|n\rangle$, é a seguinte
\begin{cases}
a_j|n\rangle = (-1)^{\sum_{k=1}^{j-1}n_k}|n'\rangle\ \text{se}\ n_j=1, \\
a_j|n\rangle = |\oslash\rangle\ \text{se}\ n_j=0, \\
\end{cases}
com $|n'\rangle$ obtido de $|n\rangle$ fazendo-se $n_j:1\rightarrow 0.$

#### Prova
Tems que
\begin{align}
a_j|n\rangle & = a_j(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = (-1)^{n_1}(a_1^\dagger)^{n_1}a_j(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = (-1)^{n_1+n_2}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}a_j\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = (-1)^{n_1+n_2+\cdots+n_{j-1}}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}a_j(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = \begin{cases} (-1)^{\sum_{k=1}^{j-1}n_k}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(I-a_{j}^\dagger a_j)(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle  \text{ se } n_j=1, \\ |\oslash\rangle \text{ se } n_j=0,  \end{cases} \\
& = \begin{cases} (-1)^{\sum_{k=1}^{j-1}n_k}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}(I-a_{j}^\dagger a_j)|vac\rangle  \text{ se } n_j=1, \\ |\oslash\rangle \text{ se } n_j=0,  \end{cases} \\
& = \begin{cases} (-1)^{\sum_{k=1}^{j-1}n_k}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle  \text{ se } n_j=1, \\ |\oslash\rangle \text{ se } n_j=0, \end{cases} \\
& = \begin{cases} (-1)^{\sum_{k=1}^{j-1}n_k}|n_1\cdots n_{j-1}0_j n_{j+1}\cdots n_n\rangle  \text{ se } n_j=1, \\ |\oslash\rangle \text{ se } n_j=0, \end{cases} \\
& = \begin{cases} (-1)^{\sum_{k=1}^{j-1}n_k}|n'\rangle  \text{ se } n_j=1, \\ |\oslash\rangle \text{ se } n_j=0, \end{cases}
\end{align}
pois 
$$(a_j^\dagger)^0 a_j|0_j\rangle = a_j|0_j\rangle = |\oslash\rangle,\ (a_j^\dagger)^1 a_j|0_j\rangle = N_j|0_j\rangle = 0|0_j\rangle = |\oslash\rangle$$
e
$$a_{j}^\dagger a_j(a_k^\dagger)^{n_k} = (-1)^{n_k}a_{j}^\dagger(a_k^\dagger)^{n_k}a_j = (-1)^{n_k+n_k}(a_k^\dagger)^{n_k}a_{j}^\dagger a_j = (a_k^\dagger)^{n_k}a_{j}^\dagger a_j.$$
$\square$

### Proposição

A ação dos operadores de levantamento em $W$, mais especificamente no estado $|n\rangle$, é a seguinte
\begin{cases}
a_j^\dagger|n\rangle = (-1)^{\sum_{k}n_k}|n'\rangle\ \text{se}\ n_j=0, \\
a_j^\dagger|n\rangle = |\oslash\rangle\ \text{se}\ n_j=1, \\
\end{cases}
com $|n'\rangle$ obtido de $|n\rangle$ fazendo-se $n_j:0\rightarrow 1.$

#### Prova
Tems que
\begin{align}
a_j^\dagger|n\rangle & = a_j^\dagger(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = (-1)^{n_1}(a_1^\dagger)^{n_1}a_j^\dagger(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = (-1)^{n_1+n_2}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}a_j^\dagger\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = (-1)^{n_1+n_2+\cdots+n_{j-1}}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}a_j^\dagger(a_{j}^\dagger)^{n_{j}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle \\
& = \begin{cases} (-1)^{\sum_{k=1}^{j-1}n_k}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}a_{j}^\dagger (a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}|vac\rangle  \text{ se } n_j=0, \\ |\oslash\rangle \text{ se } n_j=1,  \end{cases} \\
& = \begin{cases} (-1)^{\sum_{k\ne j}n_k}(a_1^\dagger)^{n_1}(a_2^\dagger)^{n_2}\cdots(a_{j-1}^\dagger)^{n_{j-1}}(a_{j+1}^\dagger)^{n_{j+1}}\cdots (a_n^\dagger)^{n_n}a_{j}^\dagger|vac\rangle  \text{ se } n_j=0, \\ |\oslash\rangle \text{ se } n_j=1,  \end{cases} \\
& = \begin{cases} (-1)^{\sum_{k\ne j}n_k}|n_1\cdots n_{j-1}1_j n_{j+1}\cdots n_n\rangle  \text{ se } n_j=0, \\ |\oslash\rangle \text{ se } n_j=1, \end{cases} \\
& = \begin{cases} (-1)^{\sum_{k\ne j}n_k}|n'\rangle  \text{ se } n_j=0, \\ |\oslash\rangle \text{ se } n_j=1, \end{cases}
\end{align}
pois 
$$(a_j^\dagger)^2 = 0_V$$
e
$$a_{j}^\dagger a_k^\dagger = -a_k^\dagger a_{j}^\dagger\ \text{pra}\ j\ne k.$$
$\square$

OBS. Esse último resultado não é igual àquele obtido por Nielsen. Além do sinal de menos, também o domínio do índice da soma é diferente. Parece que o Nielsen já colocou a somo como aparece na transformada de Jordan-Wigner.
