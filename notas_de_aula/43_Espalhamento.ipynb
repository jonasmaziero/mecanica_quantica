# Espalhamento quântico

Consideraremos o caso não relativístico de espalhamento elástico de quantons __sem spin__. O espalhamento ser __elástico__ significa que os graus de liberdade internos do quanton não mudam. Consideramos também que não há criação e/ou aniquilação de quantons pela colisão. 

Em teoria de espalhamento, fazendo incidir quantons sobre um centro espalhador e detectando os quantons espalhados, queremos descobrir as _propriedades do potencial espalhador_. Se conhecemos essas propriedades, queremos fazer previsões sobre a dependência da intensidade do feixe espalhado (probabilidade) com a direção de espalhamento. Em geral, usaremos o referencial do centro de massa, tendo assim que 
$$U(\vec{r}_{a}-\vec{r}_{b})\rightarrow U(\vec{r}),$$
como fizemos no caso do átomo de Hidrogênio.

Como o potencial não depende explicitamente do tempo, temos que 
$$\Psi(\vec{r},t)=\psi(\vec{r})e^{-iEt/\hbar}$$
e a equação a ser utilizada é a equação de Schrödinger independente do tempo:
\begin{align}
& \left(-\frac{\hbar^{2}}{2m}\nabla^{2} + U(\vec{r})\right)\psi(\vec{r})=E\psi(\vec{r})=\frac{\hbar^{2}k^{2}}{2m}\psi(\vec{r}) \\
& \therefore\ \left(\nabla^{2}+k^{2}-\frac{2m}{\hbar^{2}}U(\vec{r})\right)\psi(\vec{r})=0, \hspace{3cm} (\star)
\end{align}
em que usamos que a energia é aquela do quanton incidente livre, i.e., 
$$E=\frac{\hbar^{2}k^{2}}{2m}.$$

Lembrando do exemplo do espalhamento por um potencial barreira unimensional, tínhamos uma onda incidente $c_{ed}e^{ikz}$, uma onda refletida $c_{ee}e^{-ikz}$ e uma onda transmitida $c_{dd}e^{ikz}$. No caso tridimensional, pode-se verificar que 
\begin{equation}
\psi(\vec{r})=e^{i\vec{k}\cdot\vec{r}}
\end{equation}
satisfaz a Eq. ($\star$) quando $U(\vec{r})=0$, o que não ocorre quando $U(\vec{r})\ne0$. Isso implica que o potencial deve ter alcance finito. 

Vamos considerar uma __onda incidente__ na direção $z$, 
$$e^{ikz},$$
e teremos uma __onda espalhada na direção radial__. Pode-se verificar que, para $U=0$, a função $e^{ikr}$ não satisfaz a Eq. ($\star$). Por outro lado, usando o operador Laplaciano em coordenadas esféricas, 
$$\nabla^{2} = \frac{1}{r^{2}}\frac{\partial}{\partial r}\left(r^{2}\frac{\partial}{\partial r}\right) + \frac{1}{r^{2}\sin\theta}\frac{\partial}{\partial \theta}\left(\sin\theta\frac{\partial}{\partial \theta}\right) + \frac{1}{r^{2}\sin^{2}\theta}\frac{\partial^{2}}{\partial \phi^{2}},$$
teremos que
\begin{align}
\nabla^{2}\left(\frac{e^{ikr}}{r}\right) & = \frac{1}{r^{2}}\frac{\partial}{\partial r}\left(r^{2}\frac{\partial}{\partial r}\frac{e^{ikr}}{r}\right) + 0 + 0 \\
& = \frac{1}{r^{2}}\frac{\partial}{\partial r}\left(r^{2}\frac{ike^{ikr}r-e^{ikr}}{r^{2}}\right) \\
& = \frac{1}{r^{2}}\frac{\partial}{\partial r}\left((ikr-1)e^{ikr}\right) \\
& = \frac{1}{r^{2}}\left(ike^{ikr} + (ikr-1)ike^{ikr}\right) \\
& = -k^{2}\frac{e^{ikr}}{r}.
\end{align}
Por conseguinte, para $U(\vec{r})=0$, a função $\frac{e^{ikr}}{r}$ satisfaz a Eq. ($\star$).

No entanto, essa função não possui dependência angular. Para incluir essa dependência, usamos o seguinte __ansatz__ para a função de onda espalhada (muito longe do centro espalhador):
\begin{equation}
\psi_{e}(\vec{r}) = f_{k}(\theta,\phi)\frac{e^{ikr}}{r},
\end{equation}
em que $f_{k}(\theta,\phi)$, a chamada _amplitude de espalhamento_, é determinada por $U(\vec{r})$.

Seguindo, vamos relacionar as amplitudes de espalhamento com a seção de choque. A __seção choque diferencial__ é definida como
\begin{equation}
d\sigma = \frac{\text{No. quantons espalhados por unidade de tempo em um ângulo sólido infinitesimal }d\Omega \text{ em torno da direção }(\theta,\phi)}{\text{No. quantons incidentes por unidade de área por unidade de tempo}}.
\end{equation}

O denominador em $d\sigma$ pode ser obtido multiplicando-se a velocidade $v=p/m=\hbar k/m$ pela densidade de probabilidade $|e^{ikz}|^{2}$ ou usando a densidade de corrente de probabilidade:
\begin{align}
\vec{J} &= \frac{\hbar}{m}\Im\left(\psi^{*}\vec{\nabla}\psi\right) \\
& = \frac{\hbar}{m}\Im\left(e^{-ikz}(\hat{x}\partial_{x}+\hat{y}\partial_{y}+\hat{z}\partial_{z})e^{ikz}\right)  \\
& = \frac{\hbar}{m}\Im\left(e^{-ikz}(\hat{x}0+\hat{y}0+\hat{z}ik)e^{ikz}\right) \\
& = \frac{\hbar k}{m}\hat{z}.
\end{align}

Para obter o numerador, vamos olhar para o número de quantons em um elemento de volume $dV$ em torno do ponto $(r,\theta,\phi)$:
\begin{equation}
dN = |\psi(\vec{r})|^{2}dV = \left|f_{k}(\theta,\phi)\frac{e^{ikr}}{r}\right|^{2}r^{2}drd\Omega = |f_{k}(\theta,\phi)|^{2}drd\Omega.
\end{equation}
Como os quantons demoram um tempo $dt=dr/v$, com $v=p/m=\hbar k/m$, pra passar pelo elemento de volume $dV$, temos que o __fluxo de quantons__ é
\begin{equation}
\frac{dN}{dt} = \frac{|f_{k}(\theta,\phi)|^{2}drd\Omega}{mdr/\hbar k} = \frac{\hbar k|f_{k}(\theta,\phi)|^{2}d\Omega}{m}.
\end{equation}

Com isso, teremos que
\begin{equation}
d\sigma = \frac{\hbar k|f_{k}(\theta,\phi)|^{2}d\Omega/m}{\hbar k/m} = |f_{k}(\theta,\phi)|^{2}d\Omega.
\end{equation}

A __seção de choque__ é obtida integrando-se sobre todas as direções:
\begin{equation}
\sigma = \int d\sigma = \int |f_{k}(\theta,\phi)|^{2}d\Omega.
\end{equation}

### Análise em ondas parciais
Vamos considerar __potenciais esfericamente simétricos__, i.e., 
$$U(\vec{r}):=U(r).$$
Nesse caso, como já vimos quando da solução do átomo de Hidrogênio, a função de onda pode ser obtida usando separação de variáveis:
\begin{equation}
\psi_{n,l,m_{l}}(r,\theta,\phi) = R(r)\Upsilon_{l,m_{l}}(\theta,\phi).
\end{equation}
Quando estudamos átomos Hidrogenóides, obtivemos a seguinte equação para a parte radial da função de onda
\begin{equation}
\partial_{r}\big(r^{2}\partial_{r}R\big) + \big(r^{2}2m(E-U)/\hbar^{2}-l(l+1)\big)R=0, \hspace{2cm} (\star\star)
\end{equation}
em que usamos a notação $\partial_{r}\equiv \frac{\partial}{\partial r}$. Se definirmos
\begin{equation}
R(r) =: \frac{u(r)}{r},
\end{equation}
usando
\begin{align}
& \partial_{r}R = \partial_{r}(u/r) = \frac{(\partial_{r}u)r-u\partial_{r}r}{r^{2}} = \frac{(\partial_{r}u)r-u}{r^{2}}, \\
& \partial_{r}\big(r^{2}\partial_{r}R\big) = \partial_{r}\left(r^{2}\frac{(\partial_{r}u)r-u}{r^{2}}\right) = (\partial_{rr}u)r + (\partial_{r}u)(1) - \partial_{r} u = (\partial_{rr}u)r,
\end{align}
reescrevemos a Eq. ($\star\star$) como segue:
\begin{align}
& (\partial_{rr}u)r + \big(r^{2}2m(E-U)/\hbar^{2}-l(l+1)\big)u/r=0, \\
& \therefore\ (\partial_{rr}u) + \big(2m(E-U)/\hbar^{2}-l(l+1)/r^{2}\big)u=0, \\
& \therefore\ -\frac{\hbar^{2}}{2m}\frac{\partial^{2}u}{\partial r^{2}} + \left(U(r) + \frac{\hbar^{2}}{2m}\frac{l(l+1)}{r^{2}}\right)u = Eu.
\end{align}

Para $r\gg 1$, temos que $U=0$ e o __termo centrífugo__, 
$$ \frac{\hbar^{2}}{2m}\frac{l(l+1)}{r^{2}},$$
também pode ser desconsiderado. Com isso teremos aproximadamente que
\begin{equation}
-\frac{\hbar^{2}}{2m}\frac{\partial^{2}u}{\partial r^{2}} = Eu = \frac{\hbar^{2}k^{2}}{2m}u \ \therefore\ \frac{\partial^{2}u}{\partial r^{2}} = -k^{2}u,
\end{equation}
cuja solução geral é 
$$u(r)=c_{p}e^{ikr}+c_{n}e^{-ikr}.$$
Como o segundo termo representa uma onda chegando ao centro espalhador, fazemos $c_{n}=0$. Obtemos assim 
\begin{align}
R(r) = \frac{u(r)}{r} \approx \frac{e^{ikr}}{r},
\end{align}
como obtivemos acima através de considerações físicas.

Essa região para a qual $kr\gg 1$ é chamada de __zona de radiação__. Por outro lado, em uma região intermediária, não podemos ignorar o termo centrífugo, e teremos que resolver a seguinte equação diferencial:
\begin{align}
& -\frac{\hbar^{2}}{2m}\frac{\partial^{2}u}{\partial r^{2}} + \frac{\hbar^{2}}{2m}\frac{l(l+1)}{r^{2}}u = Eu = \frac{\hbar^{2}k^{2}}{2m}u \\
& \ \therefore\ -\frac{\partial^{2}u}{\partial r^{2}} + \frac{l(l+1)}{r^{2}}u = k^{2}u.
\end{align}

A solução desta equação diferencial é uma combinação das _funções esféricas de Bessel e de Neumann_:
\begin{equation}
u(r) = c_{j}rj_{l}(kr) + c_{n}rn_{l}(kr).
\end{equation}
Algumas __funções esféricas de Bessel__ são
\begin{align}
& j_{0}(kr) = \frac{\sin(kr)}{kr}, \\
& j_{1}(kr) = \frac{\sin(kr)}{(kr)^{2}}-\frac{\cos(kr)}{kr}, \\
& j_{2}(kr) = \left(\frac{3}{(kr)^{3}}-\frac{1}{kr}\right)\sin(kr)-\frac{3}{(kr)^{2}}\cos(kr), \\
& \vdots \\
& \text{ e } j_{l}(kr\ll 1) \approx \frac{l!2^{l}}{(2l+1)!}(kr)^{l}.
\end{align}
Algumas __funções esféricas de Neumann__ são
\begin{align}
& n_{0}(kr) = -\frac{\cos(kr)}{kr}, \\
& n_{1}(kr) = -\frac{\cos(kr)}{(kr)^{2}}-\frac{\sin(kr)}{kr}, \\
& n_{2}(kr) = -\left(\frac{3}{(kr)^{3}}-\frac{1}{kr}\right)\cos(kr)-\frac{3}{(kr)^{2}}\sin(kr) \\
& \vdots \\
& \text{ e } n_{l}(kr\ll 1) \approx \frac{(2l)!}{l!2^{l}}\frac{1}{(kr)^{l+1}}.
\end{align}

Como exemplificado acima, as funções $j_{l}$ e $n_{l}$ são combinações de senos e cossenos. Como precisamos de combinações de funções do tipo $e^{\pm ikr}$, usaremos as chamadas __funções esféricas de Hankel__:
\begin{align}
& h_{l}^{(1)}(kr) = j_{l}(kr)+in_{l}(kr), \\
& h_{l}^{(2)}(kr) = j_{l}(kr)-in_{l}(kr).
\end{align}

Algumas funções de Hankel são
\begin{align}
& h_{0}^{(1)} = -i\frac{e^{ikr}}{kr}, \\
& h_{1}^{(1)} = \left(-\frac{i}{(kr)^{2}}-\frac{1}{kr}\right)e^{ikr} \\
& h_{2}^{(1)} = \left(-\frac{3i}{(kr)^{3}}-\frac{3}{(kr)^{2}}+\frac{i}{kr}\right)e^{ikr}, \\
& \vdots \\
& h_{0}^{(2)} = i\frac{e^{-ikr}}{kr}, \\
& h_{1}^{(2)} = \left(\frac{i}{(kr)^{2}}-\frac{1}{kr}\right)e^{-ikr} \\
& h_{2}^{(2)} = \left(\frac{3i}{(kr)^{3}}-\frac{3}{(kr)^{2}}+\frac{i}{kr}\right)e^{-ikr}, \\
& \vdots.
\end{align}
Para $kr\gg 1$, temos que
\begin{align}
& h_{l}^{(1)} \approx \frac{(-i)^{l+1}e^{ikr}}{kr}, \\
& h_{l}^{(2)} \approx \frac{(i)^{l+1}e^{-ikr}}{kr}.
\end{align}

Com isso, como queremos ondas espalhadas, e pode-se escrever
\begin{equation}
u(r)=b_{1}rh_{l}^{(1)}+b_{2}rh_{l}^{(2)},
\end{equation}
fazemos a constante $b_{2}=0$. Com isso, vem que
\begin{equation}
R(r) = \frac{u(r)}{r} = b_{1}h_{l}^{(1)}(kr).
\end{equation}
Por conseguinte, podemos escrever a função de onda espalhada fora da região de espalhamento (onde $U=0$) como
\begin{equation}
\psi_{e}(r,\theta,\phi) = \sum_{l,m}c_{l,m_{l}}h_{l}^{(1)}(kr)\Upsilon_{l,m_{l}}(\theta,\phi).
\end{equation}

Lembrando
$$\Upsilon_{l,m_{l}}(\theta,\phi)=\Theta_{l,m_{l}}(\theta)\Phi_{m_{l}}(\phi).$$
Se usamos 
$$\Phi_{m_{l}}(\phi)=e^{im_{l}\phi},$$
a amplitude de espalhamento não depende do ângulo longitudinal $\phi$, o que é esperado pela __simetria esférica do potencial__. Com isso usamos
\begin{equation}
m_{l}:=0.
\end{equation}
Nesse caso 
$$\Theta_{l,m_{l}}(z)=(1-z^{2})^{|m_{l}|/2}\frac{\partial^{m_{l}}P_{l}(z)}{\partial z^{m_{l}}}\ \therefore\ \Theta_{l,0}(z)=(1-z^{2})^{0/2}\frac{\partial^{0}P_{l}(z)}{\partial z^{0}}=P_{l}(z),$$
em que $z=\cos\theta$ e $P_{l}$ são os polinômios de Legendre. Então, por convenção, usa-se as constantes $c_{l,0}$ tais que:
\begin{equation}
\psi_{e}(r,\theta,\phi) = k\sum_{l=0}^{\infty}a_{l}i^{l+1}(2l+1)h_{l}^{(1)}(kr)P_{l}(\cos\theta).
\end{equation}
Com isso, para $kr$ muito grande podemos usar $h_{l}^{(1)} \approx \frac{(-i)^{l+1}e^{ikr}}{kr}$ para escrever
\begin{align}
\psi_{e}(r,\theta,\phi) & \approx \left(\sum_{l=0}^{\infty}a_{l}(2l+1)P_{l}(\cos\theta)\right)\frac{e^{ikr}}{kr} \\
& \equiv f_{k}(\theta,\phi)\frac{e^{ikr}}{kr}.
\end{align}
Ou seja
\begin{equation}
f_{k}(\theta,\phi) = \sum_{l=0}^{\infty}a_{l}(2l+1)P_{l}(\cos\theta).
\end{equation}

Assim, a seção de choque diferencial fica escrita como
\begin{align}
d\sigma & = |f_{k}(\theta,\phi)|^{2} \\
& = \left(\sum_{l=0}^{\infty}a_{l}(2l+1)P_{l}(\cos\theta)\right)\left(\sum_{l'=0}^{\infty}a_{l'}^{*}(2l'+1)P_{l'}(\cos\theta)^{*}\right)d\Omega \\
& = \sum_{l,l'=0}^{\infty}a_{l}a_{l'}^{*}(2l'+1)(2l+1)P_{l}(\cos\theta)P_{l'}(\cos\theta)d\Omega,
\end{align}
em que usamos que $P_{l}(z)=\frac{1}{l!2^{l}}\partial_{z^{l}}(z^{2}-1)^{l}$ é um número real.

Usando 
$$\int_{0}^{\pi}\Theta_{l,m_{l}}(\theta)\Theta_{l',m_{l}}(\theta)\sin\theta d\theta = \delta_{l,l'}\frac{2}{2l+1}\frac{(l+|m_{l}|)!}{(l-|m_{l}|)!},$$
obtemos a seção de choque
\begin{align}
\sigma &= \sum_{l,l'=0}^{\infty}a_{l}a_{l'}^{*}(2l'+1)(2l+1)\iint d\Omega P_{l}(\cos\theta)P_{l'}(\cos\theta) \\
& = \sum_{l,l'=0}^{\infty}a_{l}a_{l'}^{*}(2l'+1)(2l+1)\int_{0}^{2\pi}d\phi \int_{0}^{\pi}\Theta_{l,0}(\theta)\Theta_{l',0}(\theta)\sin\theta d\theta \\
& = \sum_{l,l'=0}^{\infty}a_{l}a_{l'}^{*}(2l'+1)(2l+1)2\pi \delta_{l,l'}\frac{2}{2l+1}\frac{(l+0)!}{(l-0)!} \\
& = 4\pi\sum_{l=0}^{\infty}|a_{l}|^{2}(2l+1).
\end{align}

### Exemplo: Espalhamento quântico por uma esfera compacta
Consideremos uma energia potencial do tipo _esfera rígida_:
\begin{equation}
U(r) = \begin{cases} \infty \text{ para } r\le a, \\ 0 \text{ para } r>a. \end{cases}
\end{equation}
Temos assim a _condição de contorno_
$$\psi(a,\theta,\phi)=0,$$ ou seja, a interferência entre a onda incidente e a onda espalhada deve ser maximamente destrutiva em $r=a$:
\begin{equation}
\psi_{i}(a,\theta,\phi) = -\psi_{e}(a,\theta,\phi)\ \therefore\ e^{ika} = -k\sum_{l=0}^{\infty}a_{l}i^{l+1}(2l+1)h_{l}^{(1)}(ka)P_{l}(\cos\theta). \hspace{1cm} (\diamond)
\end{equation}

De forma geral, a expansão de uma onda plana em termos de ondas esféricas é dada pela _fórmula de Rayleigh_:
\begin{equation}
e^{ikz} = \sum_{l=0}^{\infty}i^{l}(2l+1)j_{l}(kr)P_{l}(\cos\theta).
\end{equation}
Juntando com a Eq. ($\diamond$), obtemos
\begin{align}
& \sum_{l=0}^{\infty}i^{l}(2l+1)j_{l}(ka)P_{l}(\cos\theta) + k\sum_{l=0}^{\infty}a_{l}i^{l+1}(2l+1)h_{l}^{(1)}(ka)P_{l}(\cos\theta) = 0 \\
&  \therefore\ \sum_{l=0}^{\infty}i^{l}(2l+1)P_{l}(\cos\theta)\left(j_{l}(ka) + ika_{l}h_{l}^{(1)}(ka) \right) = 0 \\
& \therefore\ a_{l} = -\frac{j_{l}(ka)}{ikh_{l}^{(1)}(ka)}.
\end{align}

Usando essa expressão para a amplitude de espalhamento, obtemos a seção de choque
\begin{equation}
\sigma = 4\pi\sum_{l=0}^{\infty}(2l+1)\left|\frac{j_{l}(ka)}{kh_{l}^{(1)}(ka)}\right|^{2}.
\end{equation}

Vamos olhar essa expressão no limite de __baixas energias__, i.e., para 
$$ka\ll 1\ \therefore\ \lambda\gg 2\pi a,$$
que significa que o comprimento de onda é muito maior que o diâmetro do centro espalhador. Como 
$$j_{l}(ka)\propto (ka)^{l} \text{ e } n_{l}(ka)\propto 1/(ka)^{l+1},$$ 
teremos que 
$$n_{l}(ka)\gg j_{l}(ka).$$
Com isso vem que
\begin{align}
\frac{j_{l}(ka)}{h_{l}^{(1)}(ka)} & = \frac{j_{l}(ka)}{j_{l}(ka)+in_{l}(ka)} \approx \frac{j_{l}(ka)}{in_{l}(ka)} \approx \frac{\frac{l!2^{l}}{(2l+1)!}(ka)^{l}}{i\frac{(2l)!}{l!2^{l}}\frac{1}{(ka)^{l+1}}} \\
& = \frac{1}{i(2l+1)}\left(\frac{l!2^{l}}{(2l)!}\right)^{2}(ka)^{2l+1}.
\end{align}
Portanto
\begin{equation}
\sigma \approx \frac{4\pi}{k^{2}}\sum_{l=0}^{\infty}\left(\frac{l!2^{l}}{(2l)!}\right)^{4}\frac{(ka)^{4l+2}}{(2l+1)}.
\end{equation}
Como $ka\ll1$, contribuições para $l>0$ são muito menores do que para $l=0$. Considerando somente este termo mais significativo, teremos
\begin{align}
\sigma \approx \frac{4\pi}{k^{2}}(1)\frac{k^{2}a^{2}}{1} = 4\pi a^{2}.
\end{align}
Esta seção de choque é independente de $\theta$, como no caso clássico. Mas no caso quântico a esfera maciça tem um tamanho efetivo maior. Enquanto que uma partícula clássica "interage" somente com a área da seção reta, $\pi a^{2}$, o quanton "interage" com toda a área da esfera, $4\pi a^{2}$.
