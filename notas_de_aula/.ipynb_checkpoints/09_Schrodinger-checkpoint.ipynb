# Equação de Schrödinger
Essa equação é uma das equações mais importantes de toda a história da humanidade. A equação de Schrödinger (EqS) é a equação diferencial que precisaremos resolver para obter a função de onda que descreve um quanton. A EqS tem a forma
\begin{equation}
\color{blue}{i\hbar\partial_{t}\Psi=-\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\Psi+U\Psi},
\end{equation}
com $\partial_{t}\equiv\frac{\partial}{\partial_{t}}$, em coordenadas cartesianas o operador Laplaciano é $$\nabla^{2}_{\vec{r}}=\partial_{xx}+\partial_{yy}+\partial_{zz}$$
em que usamos a notação 
$$U = U(\vec{r},t)$$
e
$$\Psi=\Psi(\vec{r},t)=\Psi(x,y,z,t).$$

Note __unidade imagnária__ $i=\sqrt{-1}$, que não aparece na equação de ondas clássica, faz parte da equação de Schrödinger. Por isso, em geral $$\Psi(\vec{r},t):\mathbb{R}^{4}\rightarrow\mathbb{C}.$$

Note que a EqS envolve a __derivada primeira no tempo__ e derivadas segundas no espaço, o que contrasta com a equação de ondas clássica,
\begin{equation}
v^{2}\nabla^{2}_{\vec{r}}f=\partial_{tt}f,
\end{equation}
que envolve derivadas segundas no tempo e no espaço.

A EqS, assim como a segunda lei de Newton da Mecânica Clássica e as equações de Einstein da Relatividade Geral, __não pode ser deduzida__. No entando, podemos dar uma certa plausibilidade para ela, como faremos na sequência.

A EqS é a base da Mecânica Quântica __não relativística__, e a princípio só se aplica nesse regime de movimento, ou seja, para 
$$v\ll c.$$

## Tornando a EqS plausível
Para um quanton no limite não relativístico de baixas velocidades, se usamos as __relações de de Broglie-Einstein__,
\begin{equation}
E=h\nu=\hbar\omega \text{ e } p=\frac{h}{\lambda}=\hbar k,
\end{equation}
a equação para a energia do quanton fica escrita como:
\begin{equation}
E=\frac{P^{2}}{2m}+U \text{ ou } \hbar\omega=\frac{\hbar^{2}k^{2}}{2m}+U.
\end{equation}
Se considerarmos uma onda harmônica unidimensional escrita na forma complexa
\begin{equation}
\Psi=\Psi(x,t)=Ae^{i(kx-\omega t)},
\end{equation}
então 
$$\partial_{t}\Psi=-i\omega\Psi,\ \partial_{xx}\Psi=i^{2}k^{2}\Psi=-k^{2}\Psi \text{ e } \partial_{yy}\Psi+\partial_{zz}\Psi=0.$$
Substituindo na EqS teremos
\begin{align}
& i\hbar\partial_{t}\Psi=i\hbar(-i\omega\Psi)=\hbar\omega\Psi \\
& = -\frac{\hbar^{2}}{2m}\partial_{xx}\Psi+0+0+U\Psi = \frac{\hbar^{2}}{2m}k^{2}\Psi+U\Psi,
\end{align}
que nos dá 
$$\hbar\omega = \frac{\hbar^{2}k^{2}}{2m} + U,$$
como esperado.

## EqS independente do tempo
Boa parte dos sistemas físicos que estudaremos será aproximada de forma que a energia potencial deles não depende do tempo, i.e.,
\begin{equation}
U=U(\vec{r}).
\end{equation}
Nessas condições podemos separar a EqS em duas equações diferenciais, uma envolvendo o tempo e a outra envolvendo o espaço, fazendo __separação de variáveis__:
\begin{equation}
\Psi(\vec{r},t)=\psi(\vec{r})\phi(t).
\end{equation}
Teremos assim que
\begin{align}
& i\hbar\partial_{t}(\psi(\vec{r})\phi(t)) = \psi(\vec{r})i\hbar\partial_{t}\phi(t) \\
& = -\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}(\psi(\vec{r})\phi(t))+U(\vec{r})(\psi(\vec{r})\phi(t)) \\
& = -\phi(t)\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(\vec{r})+U(\vec{r})(\psi(\vec{r})\phi(t)).
\end{align}
Dividimos essa equação por $\psi(\vec{r})\phi(t)$ para obter
\begin{align}
\frac{i\hbar\partial_{t}\phi(t)}{\phi(t)} = -\frac{1}{\psi(\vec{r})}\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(\vec{r})+U(\vec{r}).
\end{align}

Como $\vec{r}$ e $t$ são variáveis independentes, mudanças em um lado da equação acima não podem afetar o outro lado. Portanto os dois lados dessa equação tem que ser iguais a uma mesma constante $C:$
\begin{equation}
i\hbar\partial_{t}\phi(t)=C\phi(t) \therefore \partial_{t}\phi(t)=-(iC/\hbar)\phi(t) \therefore \phi(t)=e^{-iCt/\hbar}.
\end{equation}
Podemos inferir que a constante é igual a energia usando o mesmo exemplo de função de onda considerada acima:
\begin{equation}
\Psi(x,t) = Ae^{i(kx-\omega t)} = Ae^{ikx}e^{-i\omega t} =: \psi(x)\phi(t).
\end{equation}
Assim
\begin{equation}
i\hbar\partial_{t}\phi(t) = i\hbar(-i\omega)\phi(t) = \hbar\omega\phi(t) = E\phi(t)=C\phi(t).
\end{equation}
Por conseguinte
\begin{equation}
C=E \Rightarrow \phi(t)=e^{-iEt/\hbar}
\end{equation}
e
\begin{equation}
\color{blue}{-\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(\vec{r})+U(\vec{r})\psi(\vec{r}) = E\psi(\vec{r})}.
\end{equation}

Nas aplicações da Mecânica Quântica, estaremos muitas vezes preocupados em resolver essa equação, chamada de equação de Schrödinger independente do tempo (EqSIT), para algumas dependências específicas da energia potencial com a posição.

Poderíamos inferir que $C=E$ notando que a EqSIT é uma equação de autovalores e autovetores para o operador Hamiltoniano do sistema:
\begin{equation}
\left(\frac{P^{2}}{2m}+U(\vec{r})\right)\psi(\vec{r}) = H\psi(\vec{r}) = E\psi(\vec{r}),
\end{equation}
onde usamos o operador momento linear:
\begin{equation}
P \equiv \frac{\hbar}{i}\vec{\nabla}_{\vec{r}}\therefore P^{2} = -\hbar^{2}\nabla^{2}_{\vec{r}}.
\end{equation}
Veremos o porquê disso mais adiante.

## Interpretação da função de onda
A função de onda de um quanton, 
$$\Psi(\vec{r},t):\mathbb{R}^{4}\rightarrow\mathbb{C},$$
é uma função complexa. Por isso essa função não possui uma interpretação direta. Sua interpretação foi dada por Born (https://www.nobelprize.org/prizes/physics/1954/born/lecture/) e é a seguinte:
\begin{equation}
\color{blue}{|\Psi(\vec{r},t)|^{2}d^{3}r = \text{probabilidade do quanton estar no volume } d^{3}r \text{ em torno da posição } \vec{r} \text{ no instante de tempo } t.}
\end{equation}
A função de onda 
$$\Psi(\vec{r},t) \text{ é uma amplitude de probabilidade,}$$ 
e sua ligação com quantidades físicas observáveis é dada pela interpretação de Born acima.

Dada essa interpretação, a função de onda precisa satisfazer a _condição de normalização_, ou seja, se, no instante de tempo $t$, somamos as probabilidades das posições do quanton por todo o espaço, temos que obter unidade (o quanton tem que estar em algum lugar):
\begin{equation}
\iiint|\Psi(\vec{r},t)|^{2}d^{3}r=1.
\end{equation}

Para $U=U(\vec{r})$, a densidade de probabilidade também independe do tempo:
\begin{equation}
|\Psi(\vec{r},t)|^{2} = |\psi(\vec{r})\phi(t)|^{2} = |\psi(\vec{r})|^{2}|e^{-iEt/\hbar}|^{2}=|\psi(\vec{r})|^{2}.
\end{equation}

## Valores esperados
A seguir apresentamos como calcular valores médios de funções da posição e do momento linear. Veremos de onde vem as equações abaixo mais adiante.

Os valores médios de posição e do seu quadrado são calculados usando
\begin{align}
\langle \vec{R}(t)\rangle & = \iiint \vec{r}|\Psi(\vec{r},t)|^{2}d^{3}r  = \iiint \Psi(\vec{r},t)^* \vec{r}\Psi(\vec{r},t)d^{3}r \\
& = \hat{i}\iiint x|\Psi(x,y,z,t)|^{2}dxdydz + \hat{j}\iiint y|\Psi(x,y,z,t)|^{2}dxdydz + \hat{k}\iiint z|\Psi(x,y,z,t)|^{2}dxdydz \\ 
& = \hat{i}\langle X(t)\rangle + \hat{j}\langle Y(t)\rangle + \hat{k}\langle Z(t)\rangle, \\
\langle R^{2}(t)\rangle & = \iiint r^{2} |\Psi(\vec{r},t)|^{2}d^{3}r  = \iiint \Psi(\vec{r},t)^* r^2\Psi(\vec{r},t)d^{3}r \\
& = \iiint x^{2} |\Psi(x,y,z,t)|^{2}dxdydz + \iiint y^{2} |\Psi(x,y,z,t)|^{2}dxdydz + \iiint z^{2} |\Psi(x,y,z,t)|^{2}dxdydz \\ 
& = \langle X^{2}(t)\rangle + \langle Y^{2}(t)\rangle + \langle Z^{2}(t)\rangle, \\
\sigma_{X}(t) & = \sqrt{\langle R^{2}(t)\rangle-\langle \vec{R}(t)\rangle^{2}}.
\end{align}

Vale observar que todas as integrais são feitas por todo o espaço. Ademais, se mudamos de sistema de coordenadas, usamos a mesma ideia.

Para cálculos de valores médios envolvendo o momento linear usamos
\begin{align}
\langle \vec{P}(t)\rangle & = \iiint \Psi(\vec{r},t)^{*}\vec{p}\Psi(\vec{r},t)d^{3}r = \iiint \Psi(\vec{r},t)^{*}\left(\frac{\hbar}{i}\vec{\nabla}_{\vec{r}}\right)\Psi(\vec{r},t)d^{3}r \\
& = \hat{i}\iiint\Psi(x,y,z,t)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Psi(x,y,z,t)dxdydz + \hat{j}\iiint\Psi(x,y,z,t)^{*}\left(\frac{\hbar}{i}\partial_{y}\right)\Psi(x,y,z,t)dxdydz \\ 
& \hspace{0.5cm} + \hat{k}\iiint\Psi(x,y,z,t)^{*}\left(\frac{\hbar}{i}\partial_{z}\right)\Psi(x,y,z,t)dxdydz \\ 
& = \hat{i}\langle P_{x}(t)\rangle + \hat{j}\langle P_{y}(t)\rangle + \hat{k}\langle P_{z}(t)\rangle, \\
\langle P^{2}(t)\rangle & = \iiint \Psi(\vec{r},t)^{*}p^{2}\Psi(\vec{r},t)d^{3}r = \iiint \Psi(\vec{r},t)^{*}\left(-\hbar^{2}\nabla^{2}_{\vec{r}}\right)\Psi(\vec{r},t)d^{3}r \\ \\
& = \iiint\Psi(x,y,z,t)^{*}\left(-\hbar^{2}\partial_{xx}\right)\Psi(x,y,z,t)dxdydz + \iiint\Psi(x,y,z,t)^{*}\left(-\hbar^{2}\partial_{yy}\right)\Psi(x,y,z,t)dxdydz \\ 
& \hspace{0.5cm} + \iiint\Psi(x,y,z,t)^{*}\left(-\hbar^{2}\partial_{zz}\right)\Psi(x,y,z,t)dxdydz \\  
& = \langle P_{x}^{2}(t)\rangle + \langle P_{y}^{2}(t)\rangle + \langle P_{z}^{2}(t)\rangle, \\
\sigma_{P}(t) & = \sqrt{\langle P^{2}(t)\rangle-\langle \vec{P}(t)\rangle^{2}}.
\end{align}

## Sistemas bi-dimensionais 
Muitas vezes é útil, ou uma boa aproximação, considerarmos que os sistemas físicos estudados estão restritos a se movimentar em uma ou em duas dimensões. Para um sistema bi-dimensional restrito a se mover na coordenada $z=0$, podemos escrever
\begin{equation}
\Psi(x,y,z,t)\equiv \Lambda(x,y,t)\psi_{G}(z),
\end{equation}
com a função de onda na direção $z$ sendo dada por uma Gaussiana muito estreita:
\begin{equation}
\psi_{G}(z) = \frac{1}{(2\pi\sigma_z^{2})^{1/4}}e^{-(z-\mu_{z})^{2}/4\sigma_{z}^{2}} \text{ com } \mu_{z}=0 \text{ e } \sigma_{z}\ll 1.
\end{equation}
Assim, para a normalização teremos 
\begin{align}
& \iiint|\Psi(x,y,z,t)|^{2}dxdydz = \iint|\Lambda(x,y,t)|^{2}dxdy\int|\psi_{G}(z)|^{2}dz=\iint|\Lambda(x,y,t)|^{2}dxdy(1) \\
& \therefore\ \iint|\Lambda(x,y,t)|^{2}dxdy=1.
\end{align}
Para os valores médios envolvendo a posição do quanton:
\begin{align}
\langle X(t)\rangle & = \iiint x|\Lambda(x,y,t)\psi_{G}(z)|^{2}dxdydz = \iint x|\Lambda(x,y,t)|^{2}dxdy\int|\psi_{G}(z)|^{2}dz \\
& = \iint x|\Lambda(x,y,t)|^{2}dxdy.
\end{align}
Da mesma forma $\langle Y(t)\rangle = \iint y|\Lambda(x,y,t)|^{2}dxdy$, e
\begin{align}
\langle Z(t)\rangle & = \iiint z|\Lambda(x,y,t)\psi_{G}(z)|^{2}dxdydz = \iint|\Lambda(x,y,t)|^{2}dxdy\int z|\psi_{G}(z)|^{2}dz \\
& =0.
\end{align}

Para o momento linear teremos
\begin{align}
\langle P_{x}(t)\rangle & = \iiint \Lambda(x,y,t)^{*}\psi_{G}(z)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Lambda(x,y,t)\psi_{G}(z)dxdydz \\
& = \iint \Lambda(x,y,t)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Lambda(x,y,t)dxdy\int|\psi_{G}(z)|^{2}dz \\
& = \iint \Lambda(x,y,t)^{*}\left(\frac{\hbar}{i}\partial_{x}\right)\Lambda(x,y,t)dxdy.
\end{align}
Da mesma forma $\langle P_{y}(t)\rangle = \iint \Lambda(x,y,t)^{*}\left(\frac{\hbar}{i}\partial_{y}\right)\Lambda(x,y,t)dxdy$. Para
\begin{align}
\langle P_{z}(t)\rangle & = \iiint \Lambda(x,y,t)^{*}\psi_{G}(z)^{*}\left(\frac{\hbar}{i}\partial_{z}\right)\Lambda(x,y,t)\psi_{G}(z)dxdydz \\
& = \iint|\Lambda(x,y,t)|^{2}dxdy\int(\hbar/i)\psi_{G}(z)^{*}\partial_{z}\psi_{G}(z)dz \\
& \propto \iint|\Lambda(x,y,t)|^{2}dxdy\int_{-\infty}^{+\infty}z|\psi_{G}(z)|^{2}dz \\
& = 0.
\end{align}

A observação mais importante a ser deixada aqui é que se restringirmos uma coordenada do quanton a um certo valor, podemos ignorá-la nos cálculos que serão feitos usando somente $\Psi(x,y,t)$. Ficará como exercício verificar que podemos fazer isso também quando duas coordenadas são fixadas. Por exemplo, se fixamos $y$ e $z$ usaremos somente $\Psi(x,t)$ para fazer os cálculos.

## Partícula livre
Vamo discutir rapidamente o sistema físico mais simples possível, a partícula livre não relativística. Em Mecânica Quântica, a descrição desse tipo de sistema envolve alguns detalhes que devem ser tratados com cuidado para evitar problemas matemáticos e de interpretação.

A função de onda que consideramos acima,
\begin{equation}
\Psi_{d}(\vec{r},t)=Ae^{i(\vec{k}\cdot\vec{r}-\omega t)},
\end{equation}
pode ser vista como descrevendo um quanton livre,
\begin{equation}
U=0\therefore E=\frac{P^{2}}{2m}=\frac{\hbar^{2}k^{2}}{2m},
\end{equation}
que se move na diração $\vec{k}$ com velocidade $\omega/k$. Para verificar essa afirmação, consideramos o movimento dos nós da parte real dessa função, 
$$|A|\cos(\vec{k}\cdot\vec{r}-\omega t),$$
que são obtidos de $$\vec{k}\cdot\vec{r}-\omega t=n\pi/2\ \therefore\ \hat{k}\cdot\vec{r} = \frac{n\pi}{2k}+\frac{\omega}{k}t$$ com $n=1,2,3,\cdots$. Essa onda se move portanto com velocidade $\omega/k$ na direção $\hat{k}$.

Para a função de onda $\Psi_{d}(\vec{r},t)$, usando
\begin{equation}
\nabla_{\vec{r}}\Psi_{d} = (\hat{i}\partial_{x}+\hat{j}\partial_{y}+\hat{k}\partial_{z})Ae^{i(k_{x}x+k_{y}y+k_{z}z-\omega t)} = (\hat{i}ik_{x}+\hat{j}ik_{y}+\hat{k}ik_{z})\Psi_{d} = i\vec{k}\Psi_{d},
\end{equation}
obteremos o valor esperado do momento linear:
\begin{align}
\langle\vec{P}(t)\rangle & = \iiint \Psi_{d}^{*}\vec{P}\Psi_{d}d^{3}r = \iiint \Psi_{d}^{*}(\hbar/i)\vec{\nabla}_{\vec{r}}\Psi_{d}d^{3}r = \iiint \Psi_{d}^{*}(\hbar/i)(i\vec{k})\Psi_{d}d^{3}r=\hbar k\hat{k} \\
& =\sqrt{2mE}\hat{k}.
\end{align}

Vale a pena mencionar que como o momento linear de um quanton com a função de onda acima é bem definido, 
$$\hat{k}\cdot\vec{p}=\hbar k\therefore\sigma_{\hat{k}\cdot\vec{P}}=0,$$
sua posição é totalmente incerta 
$$\sigma_{\hat{k}\cdot\vec{X}}=\infty.$$
Isso implicaria em uma extensão infinita para o quanton. No entanto, mesmo quando essa aproximação é razolável, como e.g. para um próton movimentando-se para se chocar com um núcleo atômico, entre a fonte do feixe de prótons e a posição do núcleo há uma separação finita. Além disso, como 
$$|\Psi_{d}(\vec{r},t)|^{2}=|A|^{2}=\text{constante},$$
a normalização da função de onda fornece $A=0$. Para fugir desse problema de normalização, usamos o chamado __método de normalização da caixa__. Para 
$$x,y,z\in[-L,L]$$
teremos $|A|=1/(2L)^{3/2}$. Assumindo $A\in\mathbb{R}$, vem que
\begin{equation}
\Psi_{d}(\vec{r},t)=\frac{1}{(2L)^{3/2}}e^{i(\vec{k}\cdot\vec{r}-\omega t)}.
\end{equation}

Como vimos anteriormente, na realidade a solução para ter um quanton localizado em uma região finita do espaço é considerar a combinação linear de um número finito de ondas harmônicas, um __pacote de ondas__, que fará com que $\sigma_{\hat{k}\cdot\vec{P}}$ e $\sigma_{\hat{k}\cdot\vec{X}}$ sejam finitos, respeitando sempre a relação de indeterminação de Heisenberg $\sigma_{\hat{k}\cdot\vec{P}}\sigma_{\hat{k}\cdot\vec{X}}\ge\hbar$.

## Propriedades da função de onda
Para garantir que quantidades fisicamente observáveis não apresentem comportamento não físico, tais como e.g. divergências e descontinuidades, impomos algumas restrições sobre a função de onda. 

A função de onda $\Psi$ existe, é solução da equação de Schrödinger e deve satisfazer as seguintes __condições__:
1. $\Psi$ deve ser contínua, unívoca e finita (em geral).
1. A Primeiras derivadas de $\Psi$ nas coordenadas espaciais devem ser contínuas, unívocas e finitas.
1. $\Psi$ deve ir pra $0$ rápido o suficiente quando as coordenadas espaciais tendem a $\infty$, para que $|\Psi|^{2}$ permaneça finita.

A condição 1 é imposta para que a densidade de probabilidade $|\Psi|^{2}$ não mude bruscamente, para que não tenha dois valores em um mesmo ponto do espaço e para que não seja infinita. Em alguns casos a função de onda pode ser infinita, desde que seja diferenciável e normalizável. A função delta é um exemplo de tal função.

Como a EqS envolve derivadas segundas nas coordenadas espaciais, pelos mesmos motivos da condição 1, impomos a condição 2 para que as derivadas segundas sejam bem comportadas. 

A condição 3 é aplicada para que integral de normalização convirja.

A EqSIT é uma equação diferencial de segunda ordem. Ao resolvê-la aparecerão duas constantes de integração que serão determinadas usualmente especificando $\psi(\vec{r})$ e $\vec{\nabla}_{\vec{r}}\psi(\vec{r})$ em um, ou mais, pontos do espaço.

## Simetria de $U(\vec{r})$ e paridade da função de onda
A EqSIT aplicada nas posições $\vec{r}$ e $-\vec{r}$ fica:
\begin{align}
& -\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(\vec{r})+U(\vec{r})\psi(\vec{r}) = E\psi(\vec{r}), \\
& -\frac{\hbar^{2}}{2m}\nabla^{2}_{\vec{r}}\psi(-\vec{r})+U(-\vec{r})\psi(-\vec{r}) = E\psi(-\vec{r}).
\end{align}
Vamos considerar uma simetria da energia potencial do tipo
\begin{equation}
U(\vec{r})=U(-\vec{r})
\end{equation}
para escrever
\begin{align}
& \left(\nabla^{2}_{\vec{r}}+\frac{2m}{\hbar^{2}}(E-U(\vec{r}))\right)\psi(\vec{r}) = 0, \\
& \left(\nabla^{2}_{\vec{r}}+\frac{2m}{\hbar^{2}}(E-U(-\vec{r}))\right)\psi(-\vec{r})=\left(\nabla^{2}_{\vec{r}}+\frac{2m}{\hbar^{2}}(E-U(\vec{r}))\right)\psi(-\vec{r})=0.
\end{align}
Subtraindo e somando essas equações obteremos
\begin{align}
& \left(\nabla^{2}_{\vec{r}}+\frac{2m}{\hbar^{2}}(E-U(\vec{r}))\right)(\psi(\vec{r})-\psi(-\vec{r})) = 0, \\
& \left(\nabla^{2}_{\vec{r}}+\frac{2m}{\hbar^{2}}(E-U(\vec{r}))\right)(\psi(\vec{r})+\psi(-\vec{r})) = 0.
\end{align}
Para que essas equações sejam satisfeitas de forma geral, é necessário que a função de onda seja par ou ímpar, i.e.,
\begin{equation}
\psi(\vec{r})=\psi(-\vec{r}) \text{ ou } \psi(\vec{r})=-\psi(-\vec{r}) \text{, respectivamente.}
\end{equation}

Algumas observações sobre funções pares e ímpares:
* A derivada de uma função par é uma função ímpar, e.g.,
\begin{align}
\partial_{x}\psi(x,y,z) &= \lim_{\Delta x\rightarrow 0}\frac{\psi(x+\Delta x,y,z)-\psi(x,y,z)}{\Delta x} \\
& = \lim_{\Delta x\rightarrow 0}\frac{\psi(-x-\Delta x,-y,-z)-\psi(-x,-y,-z)}{\Delta x} \\ 
&= -\lim_{\Delta x\rightarrow 0}\frac{\psi(-x,-y,-z)-\psi(-x-\Delta x,-y,-z)}{\Delta x} \\
& = -\partial_{x}\psi(-x,-y,-z).
\end{align}

* O produto de duas funções pares é uma função par:
\begin{equation}
h(x)=f(x)g(x) = f(-x)g(-x) = h(-x).
\end{equation}
O produto de duas funções ímpares é uma função par:
\begin{equation}
h(x)=f(x)g(x) = (-f(-x))(-g(-x)) = f(-x)g(-x) = h(-x).
\end{equation}

* A função $f(x)=x^{n}$ é par se $n$ é par ($n=2j$ para algum $j\in\mathbb{Z}$):
\begin{align}
f(-x) & = (-x)^{n} = (-x)^{2j} = ((-x)^{j})^{2} = (-x)^{j}(-x)^{j} = (-1)^{j}x^{j}(-1)^{j}x^{j} = (-1)^{2j}x^{2j} \\
& = x^{n} = f(x).
\end{align}
