# Relações de incerteza
Vimos que a existência de observáveis incompatíveis, que não comutam, possibilita a preparação de estados de superposição que tem a eles associada uma "incerteza quântica" com relação a medidas futuras. Aqui usaremos o formalismo cinemático quântico para obter relações matemáticas gerais que fornecem limitações para o produto ou soma das "incertezas" (desvios padrões) sobre dois observáveis. Este texto é baseado, em parte, em na referência 1.

## Relação de incerteza de Heisenberg-Robertson-Schrödinger (RIHRS)
Aqui usaremos como uma das principais ferramentas a __desigualdade de Cauchy-Scharz__ (DCS):
\begin{equation}
\langle\psi|\psi\rangle\langle\phi|\phi\rangle\ge|\langle\psi|\phi\rangle|^{2},
\end{equation}
como igualdade obtida se e somente se $|\psi\rangle$ e $|\phi\rangle$ são colineares. Esse resultado é válido para qualquer par de vetores $|\psi\rangle,|\phi\rangle$ em um espaço de Hilbert $\mathcal{H}.$

Considere $A$ e $B$ dois observáveis de um sistema físico preparado no estado $|\xi\rangle$. Usaremos 
$$\langle X\rangle:=\langle X\rangle_{\xi}=\langle\xi|X|\xi\rangle$$
para denotar o valor médio de um operador qualquer $X$ e $I$ para denotar o operador identidade em $\mathcal{H}$. Vamos definir ainda os seguintes vetores:
\begin{align}
& |\psi\rangle=(A-\langle A\rangle I)|\xi\rangle, \\
& |\phi\rangle=(B-\langle B\rangle I)|\xi\rangle, \label{eq:states}
\end{align}
que substituiremos na DCS. Notemos que 
\begin{align}
\langle\psi|\psi\rangle & = ((A-\langle A\rangle I)|\xi\rangle,(A-\langle A\rangle I)|\xi\rangle) \\
& = (|\xi\rangle,(A-\langle A\rangle I)^{\dagger}(A-\langle A\rangle I)|\xi\rangle) \\
& = \langle\xi|(A-\langle A\rangle I)^{2}|\xi\rangle \\
& = \mathrm{Var}(A), \\ 
\langle\phi|\phi\rangle & = \mathrm{Var}(B),
\end{align}
com $\mathrm{Var}(X)=\langle(X-\langle X\rangle I)^{2}\rangle$
sendo a variância de $X$. 

Ademais, temos que
\begin{align}
\langle\psi|\phi\rangle &= ((A-\langle A\rangle I)|\xi\rangle,(B-\langle B\rangle I)|\xi\rangle) \\
& = (|\xi\rangle,(A-\langle A\rangle I)(B-\langle B\rangle I)|\xi\rangle) \\ 
& = (|\xi\rangle,(AB-A\langle B\rangle-\langle A\rangle B+\langle A\rangle\langle B\rangle I)|\xi\rangle) \\
& = \langle AB-\langle A\rangle\langle B\rangle I\rangle \\
& =  2^{-1}\langle\{A,B\}-2\langle A\rangle\langle B\rangle\mathbb{I}\rangle+2^{-1}\langle[A,B]\rangle \\
& = \Re(\langle\psi|\phi\rangle) + i \Im(\langle\psi|\phi\rangle) \\
& = 2^{-1}(\langle AB\rangle -\langle A\rangle\langle B\rangle + \langle BA\rangle - \langle A\rangle\langle B\rangle) + i(-i 2^{-1}\langle[A,B]\rangle)
\end{align}
onde
\begin{align}
& [A,B]=AB-BA, \\
& \{A,B\}=AB+BA
\end{align}
são, respectivamente, o comutador e anti-comutador de $A$ e $B$. Note que 
$$AB = 2^{-1}(\{A,B\}+[A,B]).$$
Como $\{A,B\}$ e $[A,B]$ são, respectivamente, operadores Hermitianos e anti-Hermitianos, seus valores médios são, respectivamente, números reais e imaginários puros. Assim, considerando que 
\begin{equation}
|\langle\psi|\phi\rangle|^{2}=(\Re(\langle\psi|\phi\rangle))^{2}+(\Im(\langle\psi|\phi\rangle))^{2},
\end{equation}
depois de algumas manipulações algébricas obteremos a RIHRS:
\begin{equation}
\color{blue}{\mathrm{Var}(A)\mathrm{Var}(B)\ge (\mathrm{CovQ}(A,B))^{2}+2^{-2}|\langle[A,B]\rangle|^{2}},\label{eq:RIHRS}
\end{equation}
onde
\begin{equation}
\mathrm{CovQ}(A,B)=2^{-1}(\mathrm{Cov}(A,B)+\mathrm{Cov}(B,A))
\end{equation}
é a covariância quântica, com 
$$\mathrm{Cov}(X,Y)=\langle XY\rangle-\langle X\rangle\langle Y\rangle$$
sendo a covariância entre os observáveis $X$ e $Y$.
Vale observar que se $A$ e $B$ são observáveis compatíveis, i.e., se
$[A,B]=0$, teremos que $\mathrm{CovQ}(A,B)=\mathrm{Cov}(A,B)$.

Vamos analizar agora o __problema da possível trivialidade__ da RIHRS, que tem por objetivo geral delimitar o quão pequeno pode ser o produto das variâncias de um par de observáveis. Para isso vamos assumir que o sistema está preparado em um estado que é autovetor do observável $A$, i.e., 
$$|\xi\rangle=|a\rangle$$
com $A|a\rangle=a|a\rangle$ e $a\in\mathbb{R}$. Neste caso pode-se verificar que 
\begin{equation}
\mathrm{Var}(A)_{|a\rangle}=\mathrm{CovQ}(A,B)_{|a\rangle}=\langle[A,B]\rangle_{|a\rangle}=0.
\end{equation}
E com isso a RIHRS nos retorna 
\begin{equation}
0\mathrm{Var}(\hat{B})\ge0.
\end{equation}
Por conseguinte, neste caso, a RIHRS não fornece informação alguma sobre a possível incompatibilidade de $A$ e $B$. O mesmo tipo de limitação se aplica se $|\xi\rangle=|b\rangle$ para $B|b\rangle=b|b\rangle.$

### Relação de incerteza de Heisenberg (RIH)
Ignorando-se o termo relacionado à covariância quântica na RIHRS e tomando a raiz quadrada da expressão que resta, obtemos a __relação de incerteza de Robertson__:
\begin{equation}
\sigma_{A}\sigma_{B} \ge 2^{-1}|\langle[A,B]\rangle|,
\end{equation}
que quando aplicada a posição e momento linear é uma das __relações de incerteza de Heisenberg__:
\begin{equation}
\sigma_{X}\sigma_{P} \ge 2^{-1}|\langle[X,P]\rangle|=\hbar/2,
\end{equation}
pois $[X,P]=i\hbar I$ (como verificaremos posteriormente).

Vale observar aqui que podemos usar a relação de incerteza de Robertson para obter uma relação que envolve a soma das variâncias:
\begin{align}
& (\sigma_{A}-\sigma_{B})^{2}\ge 0 \therefore \sigma_{A}^{2}+\sigma_{B}^{2}-2\sigma_{A}\sigma_{B}\ge 0 \\
& Var(A)+Var(B) \ge 2\sigma_{A}\sigma_{B} \ge |\langle[A,B]\rangle|.
\end{align}
No entanto, essa nova relação também sofre do problema da trivialidade.

## Relação de incerteza de Maccone-Pati (RIMP)
O problema da trivialidade da RIHRS motivou Maccone e Pati a obter novas relações de incerteza livres desse problema. A RIMP lê-se:
\begin{equation}
\color{blue}{\mathrm{Var}(A)+\mathrm{Var}(B) \ge \max\left(\frac{1}{2}|\langle\xi|(A\pm B)|\xi_{\perp}\rangle|^{2},\pm i\langle[A,B]\rangle + |\langle\xi|(A\pm iB)|\xi_{\perp}\rangle|^{2}\right)}
\end{equation}
com $|\xi_{\perp}\rangle$ sendo qualquer vetor normalizado e ortogonal a $|\xi\rangle$ e o sinal é escolhido de forma a maximizar o limite inferior.

Para provar a RIMP, começamos com a regra dos paralelogramos:
\begin{equation}
2(|||\psi\rangle||^{2}+|||\phi\rangle||^{2}) = |||\psi\rangle+|\phi\rangle||^{2} + |||\psi\rangle-|\phi\rangle||^{2}.
\end{equation}

Como fizemos no caso da RIHRS, começamos definindo 
$$|\psi\rangle=(A-\langle A\rangle I)|\xi\rangle \Rightarrow |||\psi\rangle||^{2}=\mathrm{Var}(A)$$
e 
$$|\phi\rangle=(B-\langle B\rangle I)|\xi\rangle \Rightarrow |||\phi\rangle||^{2}=\mathrm{Var}(B)$$
para escrever
\begin{align}
\mathrm{Var}(A)+\mathrm{Var}(B) & = ||\phi||^{2}+||\psi||^{2} \\
& = \frac{1}{2}\left(||(|\psi\rangle+|\phi\rangle)||^{2} + |||(\psi\rangle-|\phi\rangle)||^{2}\right) \\
& \ge  \frac{1}{2}|||\psi\rangle|\pm|\phi\rangle||^{2} \\
& = \frac{1}{2}(\langle\psi|\pm\langle\phi|)(|\psi\rangle\pm|\phi\rangle)\langle\xi_{\perp}|\xi_{\perp}\rangle \\
& \ge \frac{1}{2}|(\langle\psi|\pm\langle\phi|)|\xi_{\perp}\rangle|^{2} \\
& = \frac{1}{2}|(\langle\xi|(A-\langle A\rangle I)\pm\langle\xi|(B-\langle B\rangle I))|\xi_{\perp}\rangle|^{2} \\
& =  \frac{1}{2}|\langle\xi|(A\pm B)|\xi_{\perp}\rangle - (\langle A\rangle\pm \langle B\rangle)\langle\xi|\xi_{\perp}\rangle|^{2}  \\ 
&=  \frac{1}{2}|\langle\xi|(A\pm B)|\xi_{\perp}\rangle|^{2}.
\end{align}
Para a primeira desigualdade usamos a positividade da norma, a para a segunda desigualdade usamos a desigualdade de Cauchy-Scharz.

Para obter o segundo limite inferior, começamos notando que
\begin{align}
||\psi\pm i\phi||^{2} & = (\langle\psi|\mp i\langle\phi|)(|\psi\rangle\pm i|\phi\rangle) \\
& = ||\psi||^{2}+||\phi||^{2}\pm i(\langle\psi|\phi\rangle-\langle\phi|\psi\rangle) \\
& = ||\psi||^{2}+||\phi||^{2}\pm i2i\Im(\langle\psi|\phi\rangle).
\end{align}
Vimos que
\begin{equation}
\langle\psi|\phi\rangle=CovQ(A,B)+i(-i\langle[A,B]\rangle/2).
\end{equation}
Assim
\begin{align}
||\psi\pm i\phi||^{2} & = ||\psi||^{2}+||\phi||^{2}\pm i2i(-i\langle[A,B]\rangle/2).
\end{align}
Com isso, podemos escrever
\begin{align}
Var(A)+Var(B) & = ||\psi||^{2}+||\phi||^{2} \\
& = ||\psi\pm i\phi||^{2} \mp i\langle[A,B]\rangle \\
& = (\langle\psi|\mp i\langle\phi|)(|\psi\rangle\pm i|\phi\rangle)\langle\xi_{\perp}|\xi_{\perp}\rangle \mp i\langle[A,B]\rangle \\
& \ge |(\langle\psi|\mp i\langle\phi|)|\xi_{\perp}\rangle|^{2} \mp i\langle[A,B]\rangle \\
& = |(\langle\xi|(A-\langle A\rangle I)\mp i\langle\xi|(B-\langle B\rangle I))|\xi_{\perp}\rangle|^{2} \mp i\langle[A,B]\rangle \\
& = |\langle\xi|(A\mp iB)|\xi_{\perp}\rangle+|(\langle A\rangle\mp i\langle B\rangle)\langle\xi|\xi_{\perp}\rangle|^{2} \mp i\langle[A,B]\rangle \\
& = |\langle\xi|(A\mp iB)|\xi_{\perp}\rangle|^{2} \mp i\langle[A,B]\rangle.
\end{align}

Vale observar que esses limites inferiores na RIMP serão nulos somente se $|\xi\rangle$ for autovetor simultâneo de $A$ e de $B$.

### Exemplo: <br> Relação de incerteza entre campo elétrico (fase) e o número de fótons (intensidade) do campo eletromagnético quantizado

Um modo de oscilação do campo eletromagnético dentro de uma cavidade unidimensional, quando quantizado, é descrito pelo Hamiltoniano do oscilador harmônico quântico unidimensional:
\begin{equation}
H = \hbar\omega(N+\mathbb{I}/2),
\end{equation}
onde $N=a^{\dagger}a$ é o operador número e $a^{\dagger}$ e $a$ são os operadores de levantamento (criação) e de abaixamento (aniquilação), respectivamente. Da solução analítica para os autovalores de $H$, sabe-se que
\begin{equation}
N|n\rangle = n|n\rangle\text{, } a|n\rangle = \sqrt{n}|n-1\rangle \text{, } a^{\dagger}|n\rangle =  \sqrt{n+1}|n+1\rangle,
\end{equation}
onde $n = 0, 1, 2, \cdots, \infty$. O operador Hermitiano que descreve o campo elétrico é
\begin{equation}
E_{x} = E_{0x}\sin(kz)(a+a^{\dagger}) =: E(a+a^{\dagger})\propto \text{ quadratura } X,
\end{equation}
com $E_{0x}$ e $k$ sendo escalares reais. Vamos considerar o campo preparado no seguinte __estado__ de superposição:
\begin{equation}
|\xi\rangle = \cos\theta|0\rangle+\sin\theta|1\rangle,
\end{equation}
e vamos calcular as variâncias de $N$ e de $E_{x}$ e os limites inferiores dados pelas RIHRS e RIMP para essas quantidades.

#### Variâncias
Para o operador número:
\begin{align}
\langle N\rangle_{\xi} & = \langle\xi|N(\cos\theta|0\rangle+\sin\theta|1\rangle) \\
& = \langle\xi|(\cos\theta N|0\rangle+\sin\theta N|1\rangle) \\
& = (\cos\theta\langle 0|+\sin\theta\langle 1|)(\cos\theta (0)|0\rangle+\sin\theta (1)|1\rangle) \\ 
& = \sin^{2}\theta \\
& = \langle N^{2}\rangle_{\xi}, \\
\mathrm{Var}(N) & = \langle N^{2}\rangle_{\xi} - \langle N\rangle_{\xi}^{2} \\
& = \sin^{2}\theta\cos^{2}\theta.
\end{align}

Note que $|\xi\rangle$ não é autovetor de $E_{x}$ para nenhum valor de $\theta$ pois
\begin{equation}
(a+a^{\dagger})|\xi\rangle = \sin\theta|0\rangle + \cos\theta|1\rangle + \sqrt{2}\sin\theta|2\rangle,
\end{equation}
ou seja, $\sigma_{E_{x}}>0$. Aqui
\begin{align}
\langle E_{x}\rangle_{\xi} & = E\langle\xi|(a+a^{\dagger})|\xi\rangle \\
& = E(\cos\theta\langle 0|+\sin\theta\langle 1|)(\sin\theta|0\rangle + \cos\theta|1\rangle + \sqrt{2}\sin\theta|2\rangle) \\
& = E2\sin\theta\cos\theta \\
& = E\sin 2\theta, \\
\langle E_{x}^{2}\rangle_{\xi} & = E^{2} \langle\xi|(a+a^{\dagger})(a+a^{\dagger})|\xi\rangle = E^{2} \langle\xi|(a+a^{\dagger})(\sin\theta|0\rangle + \cos\theta|1\rangle + \sqrt{2}\sin\theta|2\rangle) \\
& = E^{2} (\cos\theta\langle 0|+\sin\theta\langle 1|)(0 + \cos\theta|0\rangle + 2\sin\theta|1\rangle + \sin\theta|1\rangle + \sqrt{2}\cos\theta|2\rangle + \sqrt{6}\sin\theta|3\rangle) \\
& = E^{2} (\cos^{2}\theta + 3\sin^{2}\theta) \\
& = E^{2} (1 + 2\sin^{2}\theta) \\
& = E^{2} (1 + 2\frac{1-\cos 2\theta}{2}) \\
& = E^{2} (2-\cos 2\theta), \\
\mathrm{Var}(E_{x}) & = \langle E_{x}^{2}\rangle_{\xi} - \langle E_{x}\rangle_{\xi}^{2} \\
& = E^{2}(2 - \cos2\theta - \sin^{2}2\theta).
\end{align}

#### Limites inferiores
Usaremos $[a,a^{\dagger}]=I$ para obter o comutador de $N$ com $E_{x}$:
\begin{align}
[N,E_{x}] & = E([a^{\dagger}a,a] + [a^{\dagger}a,a^{\dagger}]) \\
& = E(a^{\dagger}[a,a] + [a^{\dagger},a]a + a^{\dagger}[a,a^{\dagger}] + [a,a^{\dagger}]a^{\dagger}) \\
          & = E(-a + a^{\dagger}) \\
          & \propto \text{ quadratura } P.
\end{align}
O valor médio do comutadoré
\begin{align}
\langle[N,E_{x}]\rangle_{\xi} & = E\langle\xi|(-a + a^{\dagger})(\cos\theta|0\rangle+\sin\theta|1\rangle) \\
& = E(\cos\theta\langle0|+\sin\theta\langle1|)(0-\sin\theta|0\rangle+\cos\theta|1\rangle+\sqrt{2}\sin\theta|2\rangle) \\
& = E(-\sin\theta\cos\theta + \sin\theta\cos\theta) \\
& = 0.
\end{align}

Para o anticomutador, teremos
\begin{align}
\{N,E_{x}\} & = E(\{a^{\dagger}a,a\} + \{a^{\dagger}a,a^{\dagger}\}) \\
& = E(a^{\dagger}aa + aa^{\dagger}a + a^{\dagger}aa^{\dagger} + a^{\dagger}a^{\dagger}a).
\end{align}
Notemos que
\begin{align}
& a^{\dagger}aa|\xi\rangle = 0,\\
& aa^{\dagger}a|\xi\rangle = \sin\theta|0\rangle,\\
& a^{\dagger}aa^{\dagger}|\xi\rangle = \cos\theta|1\rangle + 2\sqrt{2}\sin\theta|2\rangle, \\
& a^{\dagger}a^{\dagger}a = \sqrt{2}\sin\theta|2\rangle.
\end{align}
Assim
\begin{align}
\langle\{N,E_{x}\}\rangle_{\xi} & = E\langle\xi|(a^{\dagger}aa + aa^{\dagger}a + a^{\dagger}aa^{\dagger} + a^{\dagger}a^{\dagger}a)|\xi\rangle \\
& = E(\cos\theta\langle\xi|+\sin\theta\langle1|)i(0 + \sin\theta|0\rangle + \cos\theta|1\rangle + 2\sqrt{2}\sin\theta|2\rangle + \sqrt{2}\sin\theta|2\rangle)|\xi\rangle \\
& = E(2\sin\theta\cos\theta) \\
& = E\sin2\theta.
\end{align}
Ou seja, esse é um exemplo de sistema para o qual é o valor médio do anticomutador que contribui para um limite inferior positivo na RIHRS.

A covariância quântica é dada por
\begin{align}
\mathrm{CovQ}(N,E_{x})_{\xi} & = (\mathrm{Cov}(N,E_{x})_{\xi} + \mathrm{Cov}(E_{x},N)_{\xi})/2 \\
& = \left(\langle NE_{x}\rangle_{\xi} - \langle N\rangle_{\xi}\langle E_{x}\rangle_{\xi} + \langle E_{x}N\rangle_{\xi} - \langle N\rangle_{\xi}\langle E_{x}\rangle_{\xi} \right)/2 \\
                       & = 2^{-1}\langle\{N,E_{x}\}\rangle_{\xi} - \langle N\rangle_{\xi}\langle E_{x}\rangle_{\xi} \\
                       & = 2^{-1}E\sin2\theta - \sin^{2}\theta E\sin2\theta \\
                       & = E\sin2\theta(2^{-1} - \sin^{2}\theta).
\end{align}

Assim o limite inferior para a RIHRS fica escrito como segue:
\begin{align}
L_{hrs} & = (\mathrm{CovQ}(N,E_{x}))^{2} + 2^{-2}|\langle[N,E_{x}]\rangle|^{2} \\
& = \left(E\sin2\theta(2^{-1} - \sin^{2}\theta)\right)^{2} + 0.
\end{align}
Note que o termo do comutador é nulo para este exemplo.

O limite inferior para a RIMP fica escrito como segue:
\begin{align}
L_{mp} & = \frac{1}{2}|\langle\xi_{\perp}|(N\pm E_{x})|\xi\rangle|^{2} \\
& = \frac{1}{2}|\langle\xi_{\perp}|(a^{\dagger}a\pm E(a+a^{\dagger}))(\cos\theta|0\rangle+\sin\theta|1\rangle)|^{2} \\
        & = \frac{1}{2}|\langle\xi_{\perp}|(0+\sin\theta|1\rangle\pm E(0+\sin\theta|0\rangle +\cos\theta|1\rangle+\sqrt{2}\sin\theta|2\rangle))|^{2} \\
        & = \frac{1}{2}|\langle\xi_{\perp}|((\sin\theta \pm E \cos\theta)|1\rangle \pm E(\sin\theta|0\rangle +\sqrt{2}\sin\theta|2\rangle))|^{2}.
\end{align}
A escolha mais simples para $|\xi_{\perp}\rangle$ seria $|2\rangle$. Mas nesse caso $L_{mp}=\sin^{2}\theta$ será trivial para $\theta=0,\pi$. Uma possibilidade mais geral seria
\begin{equation}
|\xi_{\perp}\rangle = \frac{1}{\sqrt{2}}\left(\sin\theta|0\rangle - \cos\theta|1\rangle + |2\rangle\right).
\end{equation}
Nesse caso teremos
\begin{align}
L_{mp} & = \frac{1}{2}\left| \pm\frac{E\sin^{2}\theta}{\sqrt{2}} -\frac{\cos\theta}{\sqrt{2}}(\sin\theta\pm E\cos\theta) \pm\frac{E\sqrt{2}\sin\theta}{\sqrt{2}} \right|^{2} \\
& = \left|\mp\frac{E\cos2\theta}{2} -\frac{\sin2\theta}{4} \pm \frac{E\sin\theta}{\sqrt{2}} \right|^{2}.
\end{align}
Ainda temos pontos de trivialidade. Poderíamos procurar por outro $|\xi_{\perp}\rangle$. No entanto, a seguir usaremos o outro limite inferior da RIMP para evitar o problema da trivialidade.

Por outro lado, podemos olhar para o outro limite inferior da RIMP. Usamos
\begin{align}
\langle\xi_{\perp}|(N\pm iE_{x})|\xi\rangle & = \langle\xi_{\perp}|(N\pm iE(a+a^{\dagger}))(\cos\theta|0\rangle+\sin\theta|1\rangle) \\
& = \langle\xi_{\perp}|(\sin\theta|1\rangle\pm iE(\sin\theta|0\rangle+\cos\theta|1\rangle+\sqrt{2}\sin\theta|2\rangle)) \\
& = \frac{\sin\theta\cos\theta}{\sqrt{2}} \pm \frac{iE}{\sqrt{2}}(\sin^{2}\theta-\cos^{2}\theta+\sqrt{2}\sin\theta) \\
& = \frac{\sin2\theta}{2\sqrt{2}} \pm \frac{iE}{\sqrt{2}}(-\cos2\theta+\sqrt{2}\sin\theta)
\end{align}
\begin{align}
\tilde{L}_{mp} & = \pm i\langle[N,E_{x}]\rangle + |\langle\xi_{\perp}|(N\pm iE_{x})|\xi\rangle|^{2} \\
& = 0 + \frac{\sin^{2}2\theta}{8} + \frac{E^{2}}{2}\left(-\cos2\theta+\sqrt{2}\sin\theta\right)^{2}.
\end{align}
Aqui usamos o mesmo $|\xi_{\perp}\rangle$ que acima. Nesse caso não há problema de trivialidade para nenhum valor de $\theta$. Abaixo mostramos esses resultados graficamente.

```python
import numpy as np
from matplotlib import pyplot as plt
import math

N = 200; dth = (np.pi-0)/N; Nm = np.zeros(N)
VarN = np.zeros(N); Em = np.zeros(N); E = 1
VarE = np.zeros(N); Lhrs = np.zeros(N); Lmp = np.zeros(N)
Lmpt = np.zeros(N); thv = np.zeros(N)
th = -dth
for j in range(0,N):
    th += dth
    thv[j] = th
    Nm[j] = (math.sin(th))**2
    VarN[j] = abs(math.sin(th)*math.cos(th))
    Em[j] = math.sin(2*th) 
    VarE[j]= math.sqrt(2-math.cos(2*th)-(math.sin(2*th))**2)
    Lhrs[j] = abs(math.sin(2*th)*(0.5-(math.sin(th))**2))
    Lmp[j] = abs(math.cos(2*th)/2 - math.sin(2*th)/4 - math.sin(th)/math.sqrt(2))
    Lmpt[j] = math.sqrt(math.sin(2*th)**2/8 + \
                   (E**2/2)*(-math.cos(2*th)+math.sqrt(2)*math.sin(th))**2)
plt.plot(thv, Lhrs, label=r'$\sqrt{L_{hrs}}/E$')
plt.plot(thv, Lmp, label=r'$\sqrt{L_{mp}}$, $E=1$')
plt.plot(thv, Lmpt, label=r'$\sqrt{\tilde{L}_{mp}}$, $E=1$')
plt.xlabel(r'$\theta$'); axes = plt.gca()
axes.set_xlim([0,np.pi]); axes.set_ylim([0.0,1.8])
plt.legend(bbox_to_anchor=(1.05, 1),loc='upper left',borderaxespad=0.)
plt.show()
```

## Tópicos relacionados
Aqui vimos as chamadas relações de incerteza de preparação, onde, para um quanton preparado em um estado qualquer $|\xi\rangle$, obtivemos limites inferiores para o produto ou soma das variâncias de dois observáveis quaisquer daquele sistema quântico. Mas relações de incerteza foram desenvolvidas em várias direções e têm múltiplas aplicações. Você aprenderá mais sobre esse assunto procurando por:

* Relações de incerteza com memórias quânticas (https://arxiv.org/abs/0909.0950);
* Relações de incerteza para medidas conjuntas (https://arxiv.org/abs/quant-ph/0310070);
* Relações de limitação entre informação adiquirida e perturbação do estado do sistema (https://arxiv.org/abs/1505.01320);
* Relações de incerteza envolvendo parâmetros que não são representados por operadores Hermitianos, tais como tempo e fase (https://arxiv.org/abs/quant-ph/0105049).
* Verificação experimental de relações de incerteza (https://arxiv.org/abs/1407.7752).
* Complementaridade quântica (https://arxiv.org/abs/2011.00723).

## Relação de incerteza energia-tempo (RIET)
O tempo é só um parâmetro em MQ, não existe um operador Hermitiano que o represente. A seguir obteremos o que se chama de RIET, clarificando a que tempo essa relação se refere. 

Consideremos a equação de Heisenberg (para observáveis que não dependem explicitamente do tempo na representação de Schrödinger),
\begin{equation}
i\hbar\partial_{t}O_{t}^{H} = [O_{t}^{H},H_{t}^{H}],
\end{equation}
e a relação de incerteza de Robertson ($\sigma_{A}\sigma_{B} \ge 2^{-1}|\langle[A,B]\rangle|$):
\begin{equation}
\sigma_{O_{t}}\sigma_{H_{t}} \ge 2^{-1}|\langle[O_{t}^{H},H_{t}^{H}]\rangle|.
\end{equation}
Teremos assim que
\begin{align}
|\langle i\hbar\partial_{t}O_{t}^{H}\rangle| & = \hbar|\partial_{t}\langle O_{t}\rangle| \\
& = |\langle[O_{t}^{H},H_{t}^{H}]\rangle| \\
& \le 2\sigma_{O_{t}}\sigma_{H_{t}}.
\end{align}
Se $O$ não é uma constante de movimento, i.e., se $\partial_{t}\langle O_{t}\rangle \ne 0$, podemos escrever a seguinte desigualdade:
\begin{equation}
\sigma_{H_{t}}\frac{\sigma_{O_{t}}}{|\partial_{t}\langle O_{t}\rangle|} \ge \frac{\hbar}{2}.
\end{equation}
Definiremos agora a seguinte escala de tempo
\begin{equation}
\Delta t_{O} := \frac{\sigma_{O_{t}}}{|\partial_{t}\langle O_{t}\rangle|},
\end{equation}
o que nos leva à RIET:
\begin{equation}
\color{blue}{\sigma_{H_{t}}\Delta t_{O} \ge \frac{\hbar}{2}}.
\end{equation}
Note que o tempo que aparece nessa expressão não é em geral aquele medido com um relógio, mas sim a razão entre a largura da distribuição de probabilidades de $O$ no tempo $t$ e a taxa de mudança do centro dessa distribuição nesse instante.

### Limite de rapidez para mudanças de estados quânticos
Consideremos o sistema inicialmente preparado no estado $|\xi\rangle$. Vamos escrever o estado evoluído como
\begin{equation}
|\xi_{t}\rangle = \cos\theta_{t}|\xi\rangle + \sin\theta_{t}|\xi_{\perp}\rangle,
\end{equation}
com $\langle\xi|\xi_{\perp}\rangle=0$. Na RIET usamos o operador Hermitiano 
$$O = |\xi\rangle\langle\xi|$$
e faremos as contas na representação de Schrödinger considerando um Hamiltoniano independente do tempo:
\begin{align}
& \langle O^{2}_{t}\rangle = \langle O_{t}\rangle = \langle\xi_{t}|\xi\rangle\langle\xi|\xi_{t}\rangle = \cos^{2}\theta_{t}, \\
& \sigma_{O_{t}} = \sqrt{\langle O^{2}_{t}\rangle - \langle O_{t}\rangle^{2}} = \sqrt{\cos^{2}\theta_{t}(1-\cos^{2}\theta_{t})} = |\sin\theta_{t}\cos\theta_{t}|, \\
& \partial_{t}\langle O_{t}\rangle = -2\cos\theta_{t}\sin\theta_{t}\partial_{t}\theta_{t}, \\
& \Delta t_{O} = \frac{|\sin\theta_{t}\cos\theta_{t}|}{|-2\cos\theta_{t}\sin\theta_{t}\partial_{t}\theta_{t}|} = \frac{1}{2|\partial_{t}\theta_{t}|}, \\
& \Delta t_{O} \ge \frac{\hbar}{2\sigma_{H_{t}}} \Rightarrow \frac{1}{2|\partial_{t}\theta_{t}|} \ge \frac{\hbar}{2\sigma_{H_{t}}} \Rightarrow \frac{\sigma_{H}}{\hbar} \ge |\partial_{t}\theta_{t}| \Rightarrow \int_{0}^{t}\frac{\sigma_{H}}{\hbar}dt' \ge \int_{0}^{t}|\partial_{t'}\theta_{t'}|dt'  \\ 
& \Rightarrow \frac{\sigma_{H}t}{\hbar} \ge \theta_{t} \Rightarrow t \ge \frac{\hbar\theta_{t}}{\sigma_{H}}.
\end{align}
Então o tempo necessário para obtermos a transição de $|\xi\rangle$ para o estado ortogonal $|\xi_{\perp}\rangle$ ($\theta_{t}=\pi/2$) é limitado por:
\begin{equation}
\color{blue}{t_{\perp} \ge \frac{\hbar\pi}{\sigma_{H}2}},
\end{equation}
ou seja, quanto maior a largura da distribuição de probabilidades do Hamiltoniano em um estado, mas rápido este estado pode mudar com o tempo. Para mais detalhes, veja e.g. (https://arxiv.org/abs/1705.08023).

# Exercícios

$1.$ Prove a desigualdade de Cauchy-Scharz, $\langle\psi|\psi\rangle\langle\phi|\phi\rangle\ge|\langle\psi|\phi\rangle|^{2}\hspace{0.2cm}\forall |\psi\rangle,|\phi\rangle\in\mathcal{H}$. Verifique que a igualdade é obtida se e somente se os vetores $|\psi\rangle$ e $|\phi\rangle$ são proporcionais.

$2.$ Para $|\phi\rangle=(B-\langle B\rangle\mathbb{I})|\xi\rangle$ e $B=B^{\dagger}$, verifique que $\langle\phi|\phi\rangle=Var(B)$.

$3.$ Verifique que se $[A,B]=\mathbb{0}$, então $CovQ(A,B)=Cov(A,B)=Cov(B,A)$.

$4.$ Verifique que o comutador de dois operadores Hermitianos $A=A^{\dagger}$ e $B=B^{\dagger}$ é um operador anti-Hermitiano, i.e., $[A,B]^{\dagger}=-[A,B]$. Verifique que os autovalores (e portanto o valor médio) desse tipo de operador são números imaginários puros.

$5.$ Verifique que se o estado no qual o sistema quântico foi preparado for um autovetor do observável $A$, i.e., $|\xi\rangle=|a\rangle$ com $A|a\rangle=a|a\rangle$, então $\mathrm{Var}(A)_{\xi}=\mathrm{CovQ}(A,B)_{\xi}=\langle[A,B]\rangle_{\xi}=0$.

$6.$ Usando que $\sigma_{j}\sigma_{k}=2i\epsilon_{jkl}\sigma_{l}$ é válido para as matrizes de Pauli, verifique que $\{\sigma_{j},\sigma_{k}\}=0$ para todo $j,k=1,2,3$ com $j\ne k$.

$7.$ Considerando o grau de liberdade de spin 1/2, calcule as variâncias e o limite inferior para o produto das variâncias dos observáveis representados por $\sigma_{x}$ e $\sigma_{z}$ para o sistema no estado $|\xi\rangle = \cos(\theta/2)|z_{+}\rangle+\sin(\theta/2)|z_{-}\rangle$. Faça o gráfico de $\mathrm{Var}(\sigma_{x})$, $\mathrm{Var}(\sigma_{z})$, $\mathrm{Var}(\sigma_{x})\mathrm{Var}(\sigma_{z})$ e do limite inferior da RIHRS para esse produto em função de $\theta\in[0,\pi]$.

$8.$ Verifique a validade geral da regra do paralelogramo $2(|||\psi\rangle||^{2}+|||\phi\rangle||^{2}) = |||\psi\rangle+|\phi\rangle||^{2} + |||\psi\rangle-|\phi\rangle||^{2}$.

$9.$ Calcule os limites inferiores na RIMP considerando o mesmo estado preparado e par de observáveis considerados no exercício 7. Discuta a vantagem de usarmos a RIMP no lugar da RIHRS nos casos em que $|\xi\rangle$ é autovetor de algum daqueles observáveis.

$10.$ Verifique que se um certo Hamiltoniano é independente do tempo, então seu desvio padrão é constante.

$11.$ Considere o Hamiltoniano $H=-2^{-1}\hbar\omega\sigma_{z}$ e o estado inicial $|x_{+}\rangle$. Mostre que o tempo mínimo para que o estado evoluído seja $|x_{-}\rangle$ é igual a $\pi/\omega$ e que o limite inferior da relação $t_{\perp}\ge \pi\hbar/2\sigma_{H}$ é $\pi/\omega$.
