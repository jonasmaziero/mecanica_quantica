# Formalismo Quântico

Até então estudamos a Velha Mecânica Quântica e a Mecânica Quântica Ondulatória de Schrödinger. Doravante investigaremos o formalismo matemático da Mecânica Quântica (MQ), a partir do qual podemos obter todos os resultados vistos anteriormente.

Em toda teoria física, buscamos descrever os __estados__ possíveis do sistema de interesse, as __mudanças__ desses estados (evolução temporal) e as quantidades __observáveis__ e suas medidas. 

Por exemplo, em Mecânica Newtoniana (MN) o estado de um sistema em um certo instante de tempo é especificado pelos valores de posição e velocidade em relação a um certo referencial inercial, e a evolução temporal é dada pela segunda lei de Newton. Os observáveis são essencialmente as posições, das quais obtemos as velocidades usando o parâmetro tempo. Está implicito em MN que os valores dos observáveis estão determinarmos antes mesmo de serem medidos, e que todos esses observáveis podem ter valores bem definidos em um determinado instante de tempo. 

Na sequência veremos os postulados da MQ para descrever estados, mudanças desses estados, observáveis e suas medidas. Veremos que a parte da medida difere muito daquela da MN, principalmente por causa da existência de observáveis que não podem ter valores bem definidos ao mesmo tempo.

É importante que você tenha um bom conhecimento de Álgebra Linear (https://github.com/jonasmaziero/algebra_linear) antes de iniciar seus estudos de MQ.

# Postulado dos estados
O estado de um sistema quântico é descrito por um vetor normalizado em um espaço de Hilbert 
$$\mathcal{H}.$$

Por simplicidade, consideraremos $\mathcal{H}$ como sendo um espaço vetorial munido de uma função produto interno 
$$\langle \cdot|\cdot \rangle:\mathcal{H}\rightarrow\mathbb{C}.$$
Formalmente, espaços desse tipo são chamados de _espaços unitários_.

Um __espaço de Hilbert__ é um espaço unitário completo. Um espaço métrico completo, um espaço de Banach, é tal que toda sequência de Cauchy de vetores converge para um limite que também está no espaço. Uma sequência de Cauchy é uma sequência cujos elementos se tornam arbitrariamente próximos um do outro conforme a sequência progride. Formalmente, $\{v_j\}_{j=0}^\infty$ é uma sequência de Cauchy se para todo $\epsilon>0$ existe um $N>0$ tal que $n,m>N\Rightarrow |v_n -v_m|<\epsilon.$

Por __sistema__ nos referimos a um __grau de liberdade__. Por exemplo, o spin (momento angular intrínseco) do elétron é um sistema, outro sistema seria sua posição, e assim por diante. Quando consideramos dois ou mais graus de liberdade como uma única entidade, a chamamos de sistema composto.

Se dividirmos o __sistema composto__ em subsistemas, seu estado é descrito por um vetor normalizado em um espaço de Hilbert obtido tomando-se o produto tensorial dos espaços individuais 
$$\mathcal{H} = \mathcal{H_{a}}\otimes \mathcal{H_{b}}\otimes\cdots.$$

# Postulado das medidas
Quantidades observáveis (mensuráveis) são representadas por __operadores lineares Hermitianos__,
$$O=O^\dagger,$$
definidos em $\mathcal{H}$. Ao medirmos um observável obteremos um dos seus possíveis valores, que são representados pelos autovalores do operador Hermitiano. 

A Mecânica Quântica (MQ), em geral, não nos possibilita prever com certeza os resultados de medidas futuras. O que a MQ nos fornece são as frequências relativas (probabilidades) com que esses resultados são obtidos, que são calculadas via a __regra de Born__:
\begin{equation}
Pr(o|\psi) = |\langle \psi |o\rangle|^{2} = |\langle o|\psi\rangle|^{2}.
\end{equation}
Essa é a probabilidade de obtermos o valor $o$ ao medirmos o observável $O$ de um sistema que está no estado $|\psi\rangle$ imediatamente antes da medida, e $|o\rangle$ é o autovetor de $O$ correspondente ao autovalor $o$.

### Repetibilidade das medidas em MQ
Mas como sabemos em qual estado $|\psi\rangle$ o sistema foi preparado? 

Essa pergunta é respondida usando a observação experimental de que ao medirmos um observável duas ou mais vezes imediatamente subsequentes, obteremos o mesmo resultado. Claro, para que isso ocorra devemos ter
\begin{equation}
Pr(o|\psi)=1 \text{, o que só é possível se } |\psi\rangle = e^{i\phi_o}|o\rangle.
\end{equation}
Então, _postulando_ que esta varificação experimental parcial seja válida de forma geral, quando dizemos que o sistema foi preparado em um certo estado, fica sub-entendido que medimos um observável desse sistema e selecionamos o autovetor correspondente a tal estado.

### Sobre operadores Hermitianos $O = O^{\dagger}$
O autovalores desse tipo de operador são números reais, e autovetores correspondentes a autovalores diferentes são ortogonais. O conjunto de autovetores de $O$ forma uma base ortogonal para $\mathcal{H}$.

### Relação de completeza
Qualquer vetor $|\psi\rangle$ do espaço vetorial $\mathcal{H}$ pode ser escrito como uma combinação linear dos vetores de uma base discreta e completa $\{|b_{j}\rangle\}_{j=1}^{d}$, i.e.,
\begin{equation}
|\psi\rangle = \sum_{j=1}^{d}\psi_{b_{j}}|b_{j}\rangle,
\end{equation}
onde $d=\dim \mathcal{H}$. Se a base $\{|b_{j}\rangle\}_{j=1}^{d}$ for ortonormal,
$$\langle b_{j}|b_{k}\rangle = \delta_{jk},$$
pode-se verificar que os coeficientes da combinação linear são dados por
\begin{equation}
\psi_{b_{j}}=\langle b_{j}|\psi\rangle.
\end{equation}

Nesse caso também podemos escrever o operador identidade em $\mathcal{H}$ como
\begin{equation}
\mathbb{I} = \sum_{j=1}^{d} |b_{j}\rangle\langle b_{j}|.
\end{equation}

Por outro lado, se temos uma base ortonormal $\{|b\rangle\}$ com incontáveis vetores, uma base contínua, a condição para ortonormalidade é 
$$\langle b|b'\rangle = \delta(b-b').$$
O vetor $|\psi\rangle\in\mathcal{H}$ também pode ser escrito em termos dessa base:
\begin{equation}
|\psi\rangle = \int  db \psi_{b}|b\rangle.
\end{equation}
De forma análoga ao caso discreto
\begin{equation}
\psi_{b}=\langle b|\psi\rangle
\end{equation}
e
\begin{equation}
\int db |b\rangle\langle b| = \mathbb{I}.
\end{equation}

### Representação matricial e representação produto externo
Se compusermos um operador $A$ com o operador identidade, nada muda em relação à ação do primeiro, i.e.,
\begin{align}
A & = \mathbb{I} A \mathbb{I} \\
& = \sum_{j=1}^{d} |b_{j}\rangle\langle b_{j}| A \sum_{k=1}^{d} |b_{k}\rangle\langle b_{k}| \\
& = \sum_{j,k=1}^{d} \langle b_{j}| A |b_{k}\rangle  |b_{j}\rangle\langle b_{k}| \\
  & = \sum_{j,k=1}^{d} A_{b_{j},b_{k}}  |b_{j}\rangle\langle b_{k}|,
\end{align}
onde 
\begin{equation}
A_{b_{j},b_{k}} = \langle b_{j}| A |b_{k}\rangle
\end{equation}
é a __representação matricial__ do operador  $A$ na base $\{|b_{j}\rangle\}_{j=1}^{d}.$
Quando um operador é escrito na forma $A=\sum_{j,k=1}^{d} A_{b_{j},b_{k}}  |b_{j}\rangle\langle b_{k}|$, dizemos que o escrevemos na __forma produto externo__.

Se a base usada na relação de completeza for contínua, obteremos
\begin{align}
A & = \mathbb{I} A \mathbb{I} \\
& = \int db |b\rangle\langle b| A \int db' |b'\rangle\langle b'| \\
& = \iint db db' \langle b| A  |b'\rangle |b\rangle\langle b'| \\
  & = \iint db db' A_{b,b'} |b\rangle\langle b'|.
\end{align}

### Valor médio de um observável na forma "sanduíche"
Considera o valor médio do observável 
$$A=\sum_{j=1}^{d} a_{j}  |a_{j}\rangle\langle a_{j}|$$
para um sistema preparado no estado $|\psi\rangle$:
\begin{align}
\langle A\rangle_{\psi} & = \sum_{j=1}^{d}a_{j}Pr(a_{j}|\psi) \\
& = \sum_{j=1}^{d}a_{j}|\langle a_{j}|\psi\rangle|^{2} \\
& = \sum_{j=1}^{d}a_{j} \langle a_{j}|\psi\rangle^{*}\langle a_{j}|\psi\rangle  \\
                        & = \sum_{j=1}^{d}a_{j} \langle \psi| a_{j}\rangle\langle a_{j}|\psi\rangle \\
                        & = \langle \psi|\left(\sum_{j=1}^{d}a_{j} | a_{j}\rangle\langle a_{j}|\right)|\psi\rangle  \\
                        & = \langle \psi|A|\psi\rangle.
\end{align}

### Amplitudes de probabilidade e normalização
Os autovetores de um observável qualquer formam uma base ortogonal para pelo menos um sub-espaço do espaço de estados do sistema. Vamos assumir aqui que essa base é completa para todo esse espaço. É frequentemente útil escrevermos o estado do sistema imediatamente antes da medida, $|\psi\rangle$, como uma combinação linear dos autovetores do observável $O$ que vai ser medido:
\begin{align}
& |\psi\rangle = \sum_{j=1}^{d} \psi_{o_{j}} |o_{j}\rangle \text{ no caso discreto} \\
& \text{ ou } |\psi\rangle = \int do \psi_{o} |o\rangle  \text{ no caso contínuo}.
\end{align}

Se colocarmos estes vetores de estado na regra de Born, veremos que a probabilidade de obsermos o valor $o_{j}$ é
\begin{align}
Pr(o_{j}|\psi) & = |\langle o_{j}|\psi\rangle|^{2} \\
& = \left|\langle o_{j}|\sum_{k=1}^{d} \psi_{o_{k}} |o_{k}\rangle\right|^{2} \\
& = \left|\sum_{k=1}^{d} \psi_{o_{k}} \langle o_{j}|o_{k}\rangle\right|^{2} \\
& = \left|\sum_{k=1}^{d} \psi_{o_{k}} \delta_{jk} \right|^{2} \\
& = \left|\psi_{o_{j}}\right|^{2}.
\end{align}
Para espectros contínuos, obteremos a densidade de probabilidade
\begin{align}
\rho(o|\psi) & = \left|\langle b|\psi\rangle\right|^{2} \\
& = \left|\langle b|\int db' \psi_{b'} |b'\rangle\right|^{2} \\
& = \left|\int db' \psi_{b'} \langle b|b'\rangle\right|^{2} \\
& = \left|\int db' \psi_{b'} \delta(b-b')\right|^{2} \\
& = \left|\psi_{b}\right|^{2}.
\end{align}
Com isso fica claro o motivo para chamarmos $\psi_{o_{j}}$ e $\psi_{o}$ de __amplitudes de probabilidade__, o módulo quadrado dessas quantidades é uma probabilidade e uma densidade de probabilidade, respectivamente. 

Aqui também podemos ver o __motivo__ para exigirmos que os vetores de estado sejam __normalizados__:
\begin{align}
||\psi||^{2} & = \langle\psi|\psi\rangle \\
& = |\psi\rangle^\dagger |\psi\rangle \\
& = \sum_{j=1}^{d} \psi_{o_{j}}^{*} \langle o_{j}| \sum_{k=1}^{d} \psi_{o_{k}} |o_{k}\rangle \\
& = \sum_{j,k=1}^{d} \psi_{o_{j}}^{*}\psi_{o_{k}} \langle o_{j}|o_{k}\rangle \\
& = \sum_{j,k=1}^{d} \psi_{o_{j}}^{*}\psi_{o_{k}} \delta_{jk} \\ 
& = \sum_{j=1}^{d} |\psi_{o_{j}}|^2 \\
& = \sum_{j=1}^{d} Pr(o_{j}|\psi) \\
& := 1.
\end{align}
e
\begin{align}
||\psi||^{2} & = \langle\psi|\psi\rangle \\
& = |\psi\rangle^\dagger |\psi\rangle \\
& = \int do \psi_{o}^{*} \langle o| \int do' \psi_{o'} |o'\rangle \\
& = \iint do do' \psi_{o}^{*}C_{o'} \langle o|o'\rangle \\
& = \iint do do' \psi_{o}^{*}\psi_{o'} \delta(o-o') \\
             & = \int do |\psi_{o}|^{2} \\
             & = \int do \rho(o|\psi) \\
             &:= 1.
\end{align}

### Pulos quânticos
Como dito acima, quando medimos um observável $O$ e obtemos um dos seus possíveis valores, preparamos o sistema no vetor de estado igual ao autovetor correspondente ao valor obtido na medida. Ilustraremos esse efeito pelas equações:
\begin{eqnarray}
|\psi\rangle & = & \sum_{j=1}^{d} \psi_{o_{j}} |o_{j}\rangle \xrightarrow[]{\text{obtém } o_{k}} |o_{k}\rangle, \\
|\psi\rangle & = & \int do \psi_{o} |o\rangle  \xrightarrow[]{\text{obtém } o'} |o'\rangle.
\end{eqnarray}
Em física clássica, estamos acostumados com mudanças contínuas de estado (de posições e velocidades, por exemplo). Na interpretação de Copenhaguem da MQ, pode ocorrer do sistema pular "instantaneamente" de uma configuração para outra que está a uma "distância finita" da primeira. Ou seja, não podemos traçar uma trajetória no espaço de estados quânticos para descrever o que ocorre durante o processo de medida.

## Incompatibilidade quântica
### __Teorema__: 
Dois operadores Hermitianos $A$ e $B$ compartilham o mesmo conjunto de autovetores se e somente se comutarem.
### Prova:
Sejam 
\begin{align}
& A = \sum_{j=1}^{d}a_{j}|\alpha_{j}\rangle\langle \alpha_{j}|,\\
& B = \sum_{j=1}^{d}b_{j}|\beta_{j}\rangle\langle \beta_{j}|,
\end{align}
com $d:=\dim\mathcal{H}$. Note que por causa da degenerescência, podemos ter e.g. $a_{1}=a_{2}$ ou e.g. $b_{2}=b_{3}=b_{4}$. 

Começamos assumindo que os dois operadores compartilham a mesma base de autovetores, i.e., 
\begin{equation}
|\alpha_{j}\rangle\equiv|\beta_{j}\rangle:=|j\rangle.
\end{equation}
Nesse caso podemos escrever as representações produto externo: $A = \sum_{j=1}^{d} a_{j}|j\rangle\langle j|$ e $B = \sum_{j=1}^{d} b_{j}|j\rangle\langle j|.$ O comutador é então
\begin{align}
[A,B] & = AB - BA \\
& = \sum_{j=1}^{d} a_{j}|j\rangle\langle j|\sum_{k=1}^{d} b_{k}|k\rangle\langle k| - \sum_{k=1}^{d} b_{k}|k\rangle\langle k|\sum_{j=1}^{d} a_{j}|j\rangle\langle j| \\
      & = \sum_{j,k=1}^{d} a_{j}b_{k}|j\rangle\langle j|k\rangle\langle k| - \sum_{k,j=1}^{d} b_{k}a_{j}|k\rangle\langle k|j\rangle\langle j| \\
      & = \sum_{j,k=1}^{d} a_{j}b_{k}|j\rangle\langle k|\delta_{jk} - \sum_{k,j=1}^{d} b_{k}a_{j}|k\rangle\langle j|\delta_{jk} \\
      & = \sum_{j=1}^{d} a_{j}b_{j}|j\rangle\langle j| - \sum_{j=1}^{d} b_{j}a_{j}|j\rangle\langle j| = 0.
\end{align}

Seguindo com a segunda parte da prova, assumimos que os operadores comutam:
\begin{equation}
[A,B] = 0 \Rightarrow AB-BA=0 \Rightarrow AB=BA.
\end{equation}
Vem assim que, para $A|\alpha_{j}\rangle=\alpha_{j}|\alpha_{j}\rangle$, e $\forall j$:
\begin{align}
A(B|\alpha_{j}\rangle) & = (AB)|\alpha_{j}\rangle \\
& = (BA)|\alpha_{j}\rangle \\
& = B(A|\alpha_{j}\rangle) \\
& = B (a_{j}|\alpha_{j}\rangle) \\
              & = a_{j}(B|\alpha_{j}\rangle). 
\end{align}
Ou seja, nesse caso devemos ter que $B|\alpha_{j}\rangle$ e $|\alpha_{j}\rangle$ são ambos autovetores de $A$ correspondentes ao mesmo autovalor $a_{j}$. Se $a_{j}$ é degenerado, existe um subespaço de $\mathcal{H}$ gerado pelos autovetores de $A$ correspondentes ao autovalor $a_{j}$; chamaremos esse subespaço de $\mathcal{H}_{j}$. Seja $\{|\alpha_{k}^{j}\rangle\}_{k=1}^{d_{j}}$ uma base qualquer de $\mathcal{H}_{j}$, com $d_{j}=\dim(\mathcal{H}_{j})$. Temos então que 
$$A|\alpha_{k}^{j}\rangle=a_{j}|\alpha_{k}^{j}\rangle$$
para $k=1,\cdots,d_{j}$. 

Além disso, $|\alpha_{j}\rangle\in\mathcal{H}_{j}$. Como $B|\alpha_{j}\rangle\in\mathcal{H}_{j}$, vemos que
\begin{equation}
B:\mathcal{H}_{j}\rightarrow \mathcal{H}_{j}.
\end{equation}
Como $B$ é um operador Hermitiano, existe uma base de $\mathcal{H}_{j}$ que o diagonaliza. Vamos assumir que essa base é $\{|\alpha_{k}^{j}\rangle\}_{k=1}^{d_{j}}$:
\begin{equation}
B|\alpha_{k}^{j}\rangle=b_{j}|\alpha_{k}^{j}\rangle
\end{equation}
com $j=1,\cdots,d_{j}$.
Demonstramos com isso que $[A,B] = 0$ implica que $A$ e $B$ compartilham a mesma base de autovetores, concluindo assim a prova do teorema nesse caso.

### Incompatibilidade e superposição quântica

Com a prova deste teorema, sabemos que se dois observáveis $A$ e $B$ de um sistema não comutam, eles não podem compartilhar todos os autovetores. Se, medindo o observável $A$, preparamos o sistema em um estado que não é autovetor do observável $B$, este vetor deve ser obrigatoriamente uma combinação linear dos autovetores de $B$ com duas ou mais amplitudes de probabilidade não nulas. Por exemplo, se $B|a_{3}\rangle$ não é proporcional a $|a_{3}\rangle$ teremos e.g.
\begin{equation}
|?\rangle \xrightarrow[\text{sai } a_{3}]{\text{mede } A} |a_{3}\rangle = \sqrt{1/4}|b_{2}\rangle +  \sqrt{3/4}|b_{5}\rangle  \xrightarrow[]{\text{mede } B} \text{sai }  b_{2} \text{ com prob. 1/4 ou } b_{5} \text{ com prob. 3/4}.
\end{equation}

# Postulado da dinâmica
As mudanças de estados em MQ são descritas por uma transformação unitária:
\begin{equation}
|\psi_{t}\rangle = U_{t}|\psi_{0}\rangle,
\end{equation}
i.e., 
$$U_{t}U_{t}^{\dagger}=U_{t}U_{t}^{\dagger}=\mathbb{I}.$$
O operador de evolução temporal sendo obtido da __equação de Schrödinger__:
\begin{equation}
i\hbar\partial_{t} U_{t} = H_{t}U_{t},
\end{equation}
com a condição inicial 
$$U_0=\mathbb{I},$$
onde $H_{t}$ é o operador Hamiltoniano do sistema no instante de tempo $t$, $\hbar$ é a constante de Planck dividida por $2\pi$ e usamos a notação $\partial_{t} = \frac{\partial}{\partial_{t}}$.

_OBS._ As duas equações acima são equivalentes à equação de Schrödinger para o estado
\begin{equation}
i\hbar\partial_{t} |\psi_{t}\rangle = H_{t}|\psi_{t}\rangle.
\end{equation}
Verificamos esse resultado como segue: 
\begin{align}
i\hbar\partial_{t} |\psi_{t}\rangle & = i\hbar\partial_{t}(U_{t}|\psi_{0}\rangle) \\
& = (i\hbar\partial_{t}U_{t})|\psi_{0}\rangle \\
& = H_{t}U_{t}|\psi_{0}\rangle \\
                                    & = H_{t}|\psi_{t}\rangle.
\end{align}

## Solução para $U_{t}$ quando $H_{t}=H$
Nesse tipo de situação, i.e., quando o Hamiltoniano não depende do tempo, podemos verificar que a seguinte forma geral é válida:
\begin{equation}
U_{t} = e^{-iHt/\hbar}.
\end{equation}
Faremos isso primeiramente usando a decomposição espectral do Hamiltoniano, 
$$H=\sum_{j}E_{j}|E_{j}\rangle\langle E_{j}|,$$
e aplicando a função de operador:
\begin{align}
i\hbar\partial_{t} U_{t} & = i\hbar\partial_{t}e^{-iHt/\hbar} \\
& = i\hbar\partial_{t}e^{\sum_{j}(-iE_{j}t/\hbar)|E_{j}\rangle\langle E_{j}|}\\ 
& = i\hbar\partial_{t}\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\ 
& = i\hbar\sum_{j}\partial_{t}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
& = i\hbar\sum_{j}(-i/\hbar)E_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
& = \sum_{j}e^{-iE_{j}t/\hbar}E_{j}|E_{j}\rangle\langle E_{j}|  \\
& = \sum_{j}e^{-iE_{j}t/\hbar}H|E_{j}\rangle\langle E_{j}| \\
& = H\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}|  \\
& = H U_{t},
\end{align}
em que usamos $H|E_j\rangle=E_j|E_j\rangle$.

## Solução para $U_{t}$ quando $[H_{t},H_{t'}]=\mathbb{0}$
Vamos verificar que neste caso
\begin{equation}
U_{t} = e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}
\end{equation}
é solução da equação de Schrödinger. Teremos
\begin{align}
i\hbar\partial_{t} U_{t} & = i\hbar\partial_{t} e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}  \\
& = i\hbar\partial_{t}\sum_{j=0}^{\infty}\frac{1}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j} \\
                         & = i\hbar\partial_{t}\left[\mathbb{I} + \sum_{j=1}^{\infty}\frac{1}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j}\right] \\
                         & = i\hbar\left[\mathbb{0} + \sum_{j=1}^{\infty}\frac{j}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j-1}\partial_{t}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)\right] \\
                         & = i\hbar\frac{-i}{\hbar}\partial_{t}\left(\int_{0}^{t}H_{t'}dt'\right)\sum_{j=1}^{\infty}\frac{j}{j!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{j-1},
\end{align}
No último passo usamos a comutatividade do Hamiltoniano em tempos diferentes. 

Seguindo, aplicando o teorema fundamental do cálculo e fazendo a mudança de variável $k=j-1$ obtemos
\begin{align}
i\hbar\partial_{t} U_{t} & = H_{t}\sum_{k=0}^{\infty}\frac{(k+1)}{(k+1)!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{k} \\
& = H_{t}\sum_{k=0}^{\infty}\frac{1}{k!}\left(-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'\right)^{k} \\ 
                         & = H_{t}e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'} \\
                         & = H_{t}U_{t}.
\end{align}

## Solução para $U_{t}$ quando $[H_{t},H_{t'}]\ne 0$
Neste caso geral, começamos reescrevendo a __equação de Schrödinger na forma integral__:
\begin{align}
& \partial_{t}U_{t}=-(i/\hbar)H_{t}U_{t}, \\
& \therefore\ \int_{0}^{t}dt'\partial_{t'}U_{t'}=U_{t}-U_{0}=U_{t}-\mathbb{I}=-\frac{i}{\hbar}\int_{0}^{t}dt'H_{t'}U_{t'}, \\
& \therefore\ U_{t}=\mathbb{I}-\frac{i}{\hbar}\int_{0}^{t}dt'H_{t'}U_{t'}.
\end{align}
Note que o operador de evolução temporal em um certo instante de tempo $t$ depende da integral dele multiplicado pelo Hamiltoniano desde $t=0$ até $t$. Mas, como essa equação vale para qualquer instante de tempo, podemos __substituir $U_{t}$ iterativamente__:
\begin{align}
U_{t} & = \mathbb{I}-\frac{i}{\hbar}\int_{0}^{t}dt'H_{t'}\left(\mathbb{I}-\frac{i}{\hbar}\int_{0}^{t'}dt''H_{t''}U_{t''}\right) \\
& = \mathbb{I} + \left(\frac{-i}{\hbar}\right)\int_{0}^{t}dt'H_{t'} + \left(\frac{-i}{\hbar}\right)^{2}\int_{0}^{t}dt'H_{t'}\int_{0}^{t'}dt''H_{t''}U_{t''} \\
& = \mathbb{I} + \left(\frac{-i}{\hbar}\right)\int_{0}^{t}dt'H_{t'} + \left(\frac{-i}{\hbar}\right)^{2}\int_{0}^{t}dt'H_{t'}\int_{0}^{t'}dt''H_{t''}  \\ 
& \hspace{0.5cm} + \left(\frac{-i}{\hbar}\right)^{3}\int_{0}^{t}dt'H_{t'}\int_{0}^{t'}dt''H_{t''}\int_{0}^{t''}dt'''H_{t'''}+\cdots \\
& = \mathbb{I} + \sum_{j=1}^{\infty}\left(\frac{-i}{\hbar}\right)^{j}\int_{0}^{t}dt_{1}H_{t_{1}}\int_{0}^{t_{1}}dt_{2}H_{t_{2}}\cdots\int_{0}^{t_{j-1}}dt_{j}H_{t_{j}} \\
& = \mathbb{I} + \sum_{j=1}^{\infty}\left(\frac{-i}{\hbar}\right)^{j}\int_{0}^{t}dt_{1}\int_{0}^{t_{1}}dt_{2}\cdots\int_{0}^{t_{j-1}}dt_{j}H_{t_{1}}H_{t_{2}}\cdots H_{t_{j}}, \hspace{1cm} (*)
\end{align}
que é chamada de __série de Dyson__. Note que 
$$0 \le t_{j}\le t_{j-1}\le \cdots\le t'''\equiv t_{3}\le t''\equiv t_{2}\le t'\equiv t_{1}\le t.$$
Ou seja, temos que ter ordenamento temporal.

#### Operador de ordenamento temporal
O operador de ordenamento temporal $\mathcal{T}$ é definido de forma que quando aplicado a um produto de operadores, este ordena esses operadores baseado na sua dependência temporal com segue:
$$\mathcal{T}\big(H_{t_1}H_{t_2}\cdots H_{t_j}\big) = H_{t_{i_1}}H_{t_{i_2}}\cdots H_{t_{i_j}}$$
com
$$t_{i_1}\ge t_{i_2} \ge \cdots \ge t_{i_j}.$$
Ou seja, $\mathcal{T}$ coloca operadores calculados em tempos maiores mais para a esquerda (eles atuam depois).

Vamos considerar a integral da série de Dyson com somente dois hamiltonianos, i.e.,
$$J_2 = \int_{0}^{t}dt_1 \int_{0}^{t_1}dt_2 H_{t_1}H_{t_2} \equiv \int_{0}^{t}dt_1 \int_{0}^{t_1}dt_2 \mathcal{T}\big(H_{t_1}H_{t_2}\big). \hspace{1cm} (\star)$$
Na sequência, vamos tirar a dependência entre as variáveis de integração, fazendo com que todos os limites de integração variem de $0$ a $t$. 

<img src='fig_T_ordering.png' width='500'>

A região de integração da integral $J_2$ está mostrada na figura acima, da esquerda. Podemos obter o mesmo resultado para integral, i.e., percorrer a mesma área de parâmetros de integração, usando:
$$J_2 = \int_{0}^{t}dt_2 \int_{t_2}^{t}dt_1 H_{t_1}H_{t_2}.$$
Se trocarmos os limites de integração, obtemos
$$J_2 = \int_{0}^{t}dt_1 \int_{t_1}^{t}dt_2 H_{t_2}H_{t_1} = \int_{0}^{t}dt_1 \int_{t_1}^{t}dt_2 \mathcal{T}\big(H_{t_1}H_{t_2}\big).  \hspace{1cm} (\star\star)$$

Assim, das equações $(\star)$ e $(\star\star)$, obtemos
\begin{align}
J_2 & = \frac{1}{2}\Big( \int_{0}^{t}dt_1 \int_{0}^{t_1}dt_2 \mathcal{T}\big(H_{t_1}H_{t_2}\big) + \int_{0}^{t}dt_1 \int_{t_1}^{t}dt_2 \mathcal{T}\big(H_{t_1}H_{t_2}\big)\Big) \\
& = \frac{1}{2}\int_{0}^{t}dt_1 \int_{0}^{t}dt_2 \mathcal{T}\big(H_{t_1}H_{t_2}\big),
\end{align}
desacoplando assim os limites de integração.

Para a integral que aparece na série de Dyson,
$$J_j = \int_{0}^{t}dt_{1}\int_{0}^{t_{1}}dt_{2}\cdots\int_{0}^{t_{j-1}}dt_{j}H_{t_{1}}H_{t_{2}}\cdots H_{t_{j}},$$
vemos que a integração é feita em uma região limitada por um hipercubo de dimensão $j$, com a região definida por
$$t_j\ge t_{j-1}\ge\cdots\ge t_2\ge t_1.$$
O volume dessa região é $1/j!$ do volume do hipercubo, o que nos permite desacoplar os limites de integração e obter
$$J_j = \frac{1}{j!}\int_{0}^{t}dt_1\int_{0}^{t}dt_2\cdots\int_{0}^{t}dt_j\mathcal{T}\big(H_{t_1}H_{t_2}\cdots H_{t_j}\big).$$

Substituindo esta expressão na __série de Dyson__, eq. ($*$), obteremos
\begin{align}
U_t & = \mathbb{I} + \sum_{j=1}^{\infty}\left(\frac{-i}{\hbar}\right)^{j}\frac{1}{j!}\int_{0}^{t}dt_1\int_{0}^{t}dt_2\cdots\int_{0}^{t}dt_j\mathcal{T}\big(H_{t_1}H_{t_2}\cdots H_{t_j}\big) \\
& =: \mathcal{T}e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}.
\end{align}
Note que esta última expressão, que é elegante e aparece com frequência na literatura, tem nela escondida toda a complicação da 1ª igualdade.

Vamos considerar o __caso particular__ em que
$$[H_t,H_{t'}]=0\ \forall\ t,t'.$$
Nesse caso o ordenamento temporal não faz diferença, i.e.,
$$\mathcal{T}\big(H_{t_1}H_{t_2}\cdots H_{t_j}\big)\equiv H_{t_1}H_{t_2}\cdots H_{t_j}.$$
Assim, podemos escrever
\begin{align}
U_t & = \mathbb{I} + \sum_{j=1}^{\infty}\left(\frac{-i}{\hbar}\right)^{j}\frac{1}{j!}\int_{0}^{t}dt_1 H_{t_1}\int_{0}^{t}dt_2 H_{t_2}\cdots\int_{0}^{t}dt_j H_{t_j} \\
& = \mathbb{I} + \sum_{j=1}^{\infty}\frac{1}{j!}\left(\frac{-i}{\hbar}\int_{0}^{t}dt' H_{t'}\right)^{j} \\
& = e^{-\frac{i}{\hbar}\int_{0}^{t}dt' H_{t'}},
\end{align}
que é o resultado particular esperado.

$1.$ Considere um vetor qualquer $|\psi\rangle$ de um espaço de Hilbert $\mathcal{H}$. Decomponha $|\psi\rangle$ em termos de uma base ortonormal $\{|b_{j}\rangle\}_{j=1}^{\dim\mathcal{H}}$ de $\mathcal{H}$. Mostre que as componentes de $|\psi\rangle$ nessa base são dadas por $\langle b_{j}|\psi\rangle$. Use esse resultado para mostrar que o operador identidade em $\mathcal{H}$ pode ser escrito em termos da base $\{|b_{j}\rangle\}_{j=1}^{\dim\mathcal{H}}$ como segue: $\mathbb{I}=\sum_{j=1}^{\dim\mathcal{H}}|b_{j}\rangle\langle b_{j}|$.

$2.$ Considere um sistema quântico de três níveis em que a ação do operador Hamiltoniano na base ortonormal $\{|1\rangle,|2\rangle,|3\rangle\}$ é

H|2\rangle=a|2\rangle+b|3\rangle, \\
H|3\rangle=a|1\rangle+b|2\rangle+a|3\rangle.
\end{cases}
Forneça a representação matricial de $H$ na base $\{|1\rangle,|2\rangle,|3\rangle\}$.

$3.$ Mostre que os autovalores de um operador Hermitiano $A$ são números reais. Verifique que dois autovetores de $A$ correspondentes a autovalores diferentes são ortogonais.

$5.$ Verifique que se $A$ for um observável com espectro contínuo, podemos escrever seu valor médio na forma: $\langle A\rangle_{\psi}=\langle\psi|A|\psi\rangle$.

$6.$ Mostre que dois operadores Hermitianos $A$ e $B$ com espectro contínuo compartilham o mesmo conjunto de autovetores se e somente se comutarem.

$7.$ Para $H_t=H$, verifique que a equação de Schrödinger $i\hbar\partial_{t}U_{t}=H_{t}U_{t}$ é satisfeita pelo operador de evolução temporal $U_{t}=e^{-iHt/\hbar}$ o expandindo em série de Taylor (e depois fazendo a derivada).

$8.$ Verifique que se o Hamiltoniano for independe do tempo, então este comuta com o operador de evolução temporal, i.e., verifique que $[U_{t},H]=[e^{-iHt/\hbar},H]=0$.

$9.$ Verifique que se o Hamiltoniano de um quanton puder ser escrito como $H_{t} = f_{t}V$, com $f_{t}$ sendo uma função escalar do tempo $t$ e $V$ um operador Hermitiano independente do tempo, então $[H_{t},H_{t'}]=\mathbb{0}$.
