# Efeito Aharonov-Bohm

Os campos elรฉtrico,
$$\vec{E}=-\vec{\nabla}\Phi-\partial_{t}\vec{A},$$ 
e magnรฉtico,
$$\vec{B}=\vec{\nabla}\times\vec{A},$$
nรฃo mudam por transformaรงรตes de calibre (TC),
\begin{cases}
\Phi\rightarrow\Phi'=\Phi-\partial_{t}\chi, \\
\vec{A}\rightarrow\vec{A}'=\vec{A}+\vec{\nabla}\chi,
\end{cases}
com $\chi=\chi(\vec{r},t)$ sendo uma funรงรฃo escalar qualquer. Isso implica na invariรขncia das leis fรญsicas (equaรงรตes de Maxwell mais forรงa de Lorentz) sob TC, o que indica que trabalhar com os potenciais $\Phi$ e $\vec{A}$ pode ser matematicamente conveniente, mas nรฃo tem importรขncia do ponto de vista Fรญsico.

No caso quรขntico, vimos, ao estudar o Efeito Stark e considerando o acoplamento mรญnimo, que a equaรงรฃo de Schrรถdinger para um quanton de carga elรฉtrica $q$ e massa $m$ รฉ invariante por TC, i.e., 
\begin{align}
& \Big(\frac{1}{2m}\big(-i\hbar\vec{\nabla} -q\vec{A}\big)^{2} + q\Phi\Big)\Psi = i\hbar\partial_{t}\Psi \\
& \hspace{3cm} \downarrow \\
& \Big(\frac{1}{2m}\big(-i\hbar\vec{\nabla} -q\vec{A}'\big)^{2} + q\Phi'\Big)\Psi' = i\hbar\partial_{t}\Psi',
\end{align}
com a funรงรฃo de onda ganhando apenas uma fase global 
$$\Psi \rightarrow \Psi' = e^{iq\chi/\hbar}\Psi.$$
Embora a equaรงรฃo de Schrรถdinger (EqS) seja invariante por TC e essa mudanรงa nรฃo mude probabilidades, Yakir Aharonov e David Bohm mostraram que TC podem afetar o comportamento quรขntico de um sistema mesmo em regiรตes do espaรงo com campos elรฉtrico e magnรฉtico nulos.

## Energia dependente de potencias onde os campos sรฃo nulos
Primeiramente, vamos considerar um exemplo para o qual as energias podem depender dos potenciais em uma regiรฃo de campos elรฉtrico e magnรฉtico nulos. Consideremos um quanton de massa $m$ e carga elรฉtrica $q$ restrito a se __mover em um cรญrculo__ de raio $b$. Concรชntrico a este cรญrculo estรก um solenรณide de raio $a<b$. Por este solenรณide passa uma corrente elรฉtrica $I$. O solenรณide รฉ muito longo, assim o campo magnรฉtico interno รฉ uniforme, $B$, e o campo magnรฉtico externo รฉ nulo
$$\vec{B}(r>a)=0.$$
Como o solenรณide รฉ neutro, e $I$ รฉ constante, o campo elรฉtrico รฉ nulo tanto no interior quanto fora do solenรณide
$$\vec{E}(r)=0\ \forall r.$$
Se adotamos a condiรงรฃo de calibre 
$$\vec{\nabla}\cdot\vec{A}=0,$$
o potencial vetor magnรฉtico fora do solenรณide pode ser escrito como
$$\vec{A}(r>a) = \frac{\varphi}{2\pi r}\hat{\phi},$$
em que 
$$\varphi=\pi a^{2}B$$
รฉ o __fluxo magnรฉtico__ no interior do solenรณide. Pela restriรงรฃo no movimeto do quanton, 
$$r=b \text{ e } \theta=\pi/2,$$ 
a funรงรฃo de onda depende somente da variรกvel $\phi$. Neste caso 
$$\vec{\nabla} \rightarrow \hat{\phi}\frac{1}{r\sin\theta}\partial_\phi.$$ 

Com isso, a equaรงรฃo de Schrรถdinger independente do tempo fica escrita como
\begin{align}
& \frac{1}{2m}\big(-i\hbar\vec{\nabla} -q\vec{A}\big)^{2}\psi = E\psi, \\
& \therefore\ \frac{1}{2m}\big(-\hbar^{2}\nabla^{2} + 2i\hbar q\vec{A}\cdot\vec{\nabla}+q^{2}A^{2}\big)\psi = E\psi, \\
& \therefore\ \frac{1}{2m}\Big(-\frac{\hbar^{2}}{r^{2}\sin^{2}\theta}\partial_{\phi\phi} + 2i\hbar q\frac{\varphi}{2\pi r^{2}\sin\theta}\partial_{\phi} + q^{2}\frac{\varphi^{2}}{4\pi^{2}r^{2}}\Big)\psi = E\psi, \\
& \therefore\ \Big(-\frac{\hbar^{2}}{b^{2}}\partial_{\phi\phi} + \frac{i\hbar q\varphi}{\pi b^{2}}\partial_{\phi} + \frac{q^{2}\varphi^{2}}{4\pi^{2}b^{2}}\Big)\psi = 2mE\psi, \\
& \therefore\ \Big(\partial_{\phi\phi} - \frac{iq\varphi}{\hbar\pi}\partial_{\phi} + \frac{2mEb^{2}}{\hbar^{2}} - \frac{q^{2}\varphi^{2}}{4\pi^{2}\hbar^{2}} \Big)\psi=0 \\
&\therefore\ \big(\partial_{\phi\phi} - 2i\beta\partial_{\phi} + \varepsilon \Big)\psi=0,
\end{align}
em que definimos
$$\beta = \frac{q\varphi}{2\hbar\pi} \text{ e } \varepsilon=\frac{2mEb^{2}}{\hbar^{2}}-\beta^{2}.$$

```python
from sympy import symbols, Function, dsolve, Derivative
be, ep, ph = symbols('beta varepsilon phi')
ps = Function('psi')
dsolve(Derivative(ps(ph),ph,ph) - 2*1j*be*Derivative(ps(ph),ph) + ep*ps(ph), ps(ph))
```

Entรฃo, as soluรงรตes da equaรงรฃo de Schrรถdinger tem a forma geral
$$\psi(\phi) = c_{+}e^{i\lambda_{+}\phi} + c_{-}e^{i\lambda_{-}\phi}$$
com
$$\lambda_{\pm} = \beta \pm\sqrt{\beta^{2}+\varepsilon}.$$
Usaremos a soluรงรฃo particular
$$\psi(\phi) = c_{\pm}e^{i\lambda_{\pm}\phi}.$$
Se impomos a condiรงรฃo de contorno
$$\psi(\phi)=\psi(\phi+2\pi),$$
vemos que $\lambda_{\pm}$ deve ser um nรบmero inteiro, que chamaremos de 
$$n=\cdots,-2,-1,0,1,2,\cdots.$$
Assim
\begin{align}
& \lambda_{\pm} = \beta \pm\sqrt{\beta^{2}+\varepsilon} = n\\
& \therefore\ \beta^{2}+\varepsilon = \big(n-\beta\big)^{2} \\
& \therefore\ \varepsilon = \frac{2mEb^{2}}{\hbar^{2}}-\beta^{2} = \big(n-\beta\big)^{2} -\beta^{2} \\
& \therefore\ E = E_{n} = \frac{\hbar^{2}}{2mb^{2}}\big(n-\beta\big)^{2} =  \frac{\hbar^{2}}{2mb^{2}}\Big(n-\frac{q\varphi}{2\hbar\pi}\Big)^{2}.
\end{align}
Ou seja, a energia do quanton fora do solenรณide, onde o campo magnรฉtico รฉ nulo, depende do fluxo de campo magnรฉtico dentro do solenรณide. Esse fato pode ser chamado de __efeito Aharonov-Bohm__.

O operador momento linear em coordenadas esfรฉricas fica escrito como
\begin{align}
\vec{P} = & -i\hbar\vec{\nabla} \\
& = -i\hbar\hat{r}\partial_{r} -i\hbar\hat{\theta}r^{-1}\partial_{\theta} -i\hbar\hat{\phi}r^{-1}\sin^{-1}\theta\partial_{\phi} \\
& = \hat{r}P_{r} + \hat{\theta}P_{\theta} + \hat{\phi}P_{\phi}.
\end{align}
Como o quanton estรก restrito a se movimentar no cรญrculo, $P_{r}=P_{\theta}=0$. Vemos assim que
\begin{align}
P_{\phi}\psi(\phi) = & -i\hbar r^{-1}\sin^{-1}\theta i\lambda_{\pm}\psi(\phi) \\
& = \hbar b^{-1} n\psi(\phi) \\
& = p_\phi\psi(\phi).
\end{align}
Portanto, estados com $n$ positivo correspondem ao movimento do quanton no mesmo __sentido da corrente elรฉtrica__, que por sua vez estรฃo associados a um menor valor de energia. Estados com $n$ negativo correspondem ao movimento do quanton no sentido contrรกrio ao da corrente elรฉtrica e tem maior energia.

## Experimento de Aharonov e Bohm
Vamos supor que o quanton se move em uma regiรฃo de campo magnรฉtico nulo,
$$\vec{B}=\vec{\nabla}\times\vec{A}=\vec{0},$$
mas com potencial vetor diferente de zero, $$\vec{A}\ne\vec{0}.$$
A equaรงรฃo de Schrรถdinger (EqS),
$$\Big(\frac{1}{2m}\big(-i\hbar\vec{\nabla} -q\vec{A}\big)^{2} + q\Phi\Big)\Psi = i\hbar\partial_{t}\Psi,$$
pode ser simplificada se definimos
$$\Psi:=e^{ig}\Xi$$
com 
$$g=\frac{q}{\hbar}\int_{\vec{\mathcal{O}}}^{\vec{r}}\vec{A}(\vec{r}')\cdot d\vec{r}',$$
que รฉ independente do caminho, uma vez que $\vec{\nabla}\times\vec{A}=\vec{0}$. O ponto $\vec{\mathcal{O}}$ รฉ um ponto de referรชncia qualquer.

Pode-se verificar que 
$$\vec{\nabla}g= \frac{q\vec{A}(\vec{r})}{\hbar}.$$
Com isso, usando
\begin{align}
\vec{\nabla}\Psi & = \vec{\nabla}\big(e^{ig}\Xi\big) \\
& = i(\vec{\nabla}g)e^{ig}\Xi + e^{ig}\vec{\nabla}\Xi \\
& = iq\hbar^{-1}\vec{A}e^{ig}\Xi + e^{ig}\vec{\nabla}\Xi,
\end{align}
podemos escrever
\begin{align}
\big(-i\hbar\vec{\nabla}-q\vec{A}\big)\Psi =& -i\hbar iq\hbar^{-1}\vec{A}e^{ig}\Xi -i\hbar e^{ig}\vec{\nabla}\Xi -q\vec{A}e^{ig}\Xi \\
& = -i\hbar e^{ig}\vec{\nabla}\Xi.
\end{align}
Ou seja, usando 
$$\vec{\nabla}\cdot(f\vec{\nabla}g) = (\vec{\nabla}f)\cdot(\vec{\nabla}g)+f\nabla^{2}g,$$
teremos que
\begin{align}
\big(-i\hbar\vec{\nabla}-q\vec{A}\big)^{2}\Psi & = \big(-i\hbar\vec{\nabla}-q\vec{A}\big)\cdot\big(-i\hbar\vec{\nabla}-q\vec{A}\big)\Psi \\
& = \big(-i\hbar\vec{\nabla}-q\vec{A}\big)\cdot\big(-i\hbar e^{ig}\vec{\nabla}\Xi\big) \\
& = -\hbar^{2}\vec{\nabla}\cdot\big(e^{ig}\vec{\nabla}\Xi\big) + iq\hbar e^{ig}\vec{A}\cdot\vec{\nabla}\Xi \\
& = -\hbar^{2}\big(\vec{\nabla}e^{ig}\big)\cdot\big(\vec{\nabla}\Xi\big) -\hbar^{2}e^{ig}\nabla^{2}\Xi + iq\hbar e^{ig}\vec{A}\cdot\vec{\nabla}\Xi \\
& = -\hbar^{2}i(\vec{\nabla}g)e^{ig}\cdot\vec{\nabla}\Xi -\hbar^{2}e^{ig}\nabla^{2}\Xi + iq\hbar e^{ig}\vec{A}\cdot\vec{\nabla}\Xi \\
& = -\hbar^{2}iq\hbar^{-1}\vec{A}e^{ig}\cdot\vec{\nabla}\Xi -\hbar^{2}e^{ig}\nabla^{2}\Xi + iq\hbar e^{ig}\vec{A}\cdot\vec{\nabla}\Xi \\
& =-\hbar^{2}e^{ig}\nabla^{2}\Xi.
\end{align}

Com isso, a equaรงรฃo de Schrรถdinger fica escrita como segue
\begin{align}
& -\frac{\hbar^{2}}{2m}e^{ig}\nabla^{2}\Xi + q\Phi e^{ig}\Xi = i\hbar\partial_{t}(e^{ig}\Xi)  = e^{ig}i\hbar\partial_{t}\Xi \\
& \therefore\ -\frac{\hbar^{2}}{2m}\nabla^{2}\Xi + q\Phi \Xi = i\hbar\partial_{t}\Xi.
\end{align}
Vemos assim que $\Xi$ satisfaz a EqS sem $\vec{A}$. Se resolvemos essa equaรงรฃo, podemos obter a funรงรฃo de onda na presenรงa de $A$: $\Psi=e^{ig}\Xi$.  

Feitas essas consideraรงรตes, passamos para o experimento de pensamento de Aharonov-Bohm (AB). Eles consideraram um elรฉtron que รฉ preparado em um estado de superposiรงรฃo de caminhos:
$$\Psi = 2^{-1/2}(\Psi_{+}+\Psi_{-}).$$
Esses dois caminhos sรฃo aqueles em que o elรฉtron passa em "lados" opostos na parte exterior de um solenรณide, como aquele que vimos no รบltimo exemplo. O potencial vetor รฉ um campo vetorial com sentido acompanhando aquele da corrente elรฉtrica no solenรณide. Com isso teremos 
\begin{align}
g_{\pm} & = \frac{q}{\hbar}\int_{\mathcal{O}}^{r}\vec{A}(\vec{r}')\cdot d\vec{r}' \\
& = \frac{q}{\hbar}\int_{\mathcal{O}}^{r}\frac{\varphi}{2\pi r'}\hat{\phi}'\cdot r'd\phi'(\pm\hat{\phi}') \\
& = \pm\frac{q\varphi}{\hbar 2\pi}\int_{\mathcal{O}}^{r}d\phi' \\
& = \pm\frac{q\varphi}{\hbar 2\pi}\pi \\
& = \pm\frac{q\varphi}{2\hbar},
\end{align}
em que usamos $\vec{A}(r>a) = \frac{\varphi}{2\pi r}\hat{\phi}$.

Entรฃo, fazendo os caminhos se cruzarem depois de passar pelo solenรณide, o estado do elรฉtron serรก
$$\Psi = 2^{-1/2}(e^{ig_{+}}\Psi_{+}+e^{ig_{-}}\Psi_{-}),$$
com os estados dos diferentes caminhos ganhando uma diferenรงa de fase
$$\delta g=g_{+}-g_{-}=\frac{q\varphi}{\hbar},$$
que pode ser controlada variando-se a corrente elรฉtrica no solenรณide. Essa diferenรงa de fase pode causar interferรชncia destrutiva, mudando as probabilidades de detecรงรฃo do elรฉtron. Esse efeito foi medido experimentalmente em (https://doi.org/10.1103/PhysRevLett.5.3) e em (https://doi.org/10.1103/PhysRevA.34.815).

## Efeito Aharonov-Bohm como exemplo de fase de Berry
Consideremos que uma partรญcula de carga elรฉtrica $q$ e massa $m$ estรก confinada a se mover em uma pequena caixa com posiรงรฃo $\vec{R}$ em relaรงรฃo a um ponto $O$ no mesmo solenรณide considerado acima. A posiรงรฃo da partรญcula em relaรงรฃo a $O$ รฉ $\vec{r}$. A equaรงรฃo de Schrรถdinger independente do tempo fica escrita, neste caso, como segue
$$\Big(\frac{1}{2m}\big(-i\hbar\vec{\nabla}-q\vec{A}(\vec{r})\big)^{2}+q\Phi(\vec{r}-\vec{R})\Big)\psi=E\psi.$$
Como acima, fazemos
$$\psi = e^{ig}\xi$$
com $\psi=\psi(\vec{r}-\vec{R})$, $\xi=\xi(\vec{r}-\vec{R})$ e
$$g=\frac{q}{\hbar}\int_{\vec{R}}^{\vec{r}}\vec{A}(\vec{r}')\cdot d\vec{r}'$$
com $\xi$ satisfazendo a EqS:
$$\Big(-\frac{\hbar^{2}}{2m}\nabla^{2}+q\Phi(\vec{r}-\vec{R})\Big)\xi=E\xi.$$

Seguindo, consideramos que a caixa รฉ carregada em torno do solenรณide de forma a satisfazer a aproximaรงรฃo adiabรกtica. Teremos que
\begin{align}
\vec{\mathcal{A}} & = i\langle\psi|\vec{\nabla}_{\vec{R}}|\psi\rangle \\
& = i\int \psi^{*}\vec{\nabla}_{\vec{R}}\psi d^{3}r\\
& = i\int e^{-ig}\xi^{*}\vec{\nabla}_{\vec{R}}(e^{ig}\xi) d^{3}r \\
& = -i\int e^{-ig}\xi^{*}\vec{\nabla}_{\vec{r}}(e^{ig}\xi) d^{3}r \\
& = -i\int e^{-ig}\xi^{*}\big(e^{ig}i(\vec{\nabla}_{\vec{r}}g)\xi + e^{ig}\vec{\nabla}_{\vec{r}}\xi\big) d^{3}r \\
& = -i\int e^{-ig}\xi^{*}\big(e^{ig}iq\vec{A}(\vec{R})\hbar^{-1}\xi + e^{ig}(i\hbar^{-1})(-i\hbar\vec{\nabla}_{\vec{r}})\xi\big) d^{3}r \\
& = q\hbar^{-1}\vec{A}\int \xi^{*}\xi d^{3}r +\hbar^{-1}\int\xi^{*}\vec{P}\xi d^{3}r \\
& = q\hbar^{-1}\vec{A} + \vec{0},
\end{align}
em usamos $\vec{\nabla}_{\vec{R}}=-\vec{\nabla}_{\vec{r}}$ ao atuar em $\xi(\vec{r}-\vec{R})$ que o valor mรฉdio do momento linear para um estado estacionรกrio รฉ nulo. Assim a fase de Berry fica escrita como
\begin{align}
\gamma & = \oint \vec{\mathcal{A}}(\vec{R})\cdot d\vec{R} \\
& = \frac{q}{\hbar}\oint \vec{A}(\vec{R})\cdot d\vec{R} \\
& = \frac{q}{\hbar}\int_{S}\vec{\nabla}\times\vec{A}\cdot d\vec{a} \\
& = \frac{q}{\hbar}\int_{S}\vec{B}\cdot d\vec{a} \\
& = \frac{q\varphi}{\hbar}.
\end{align}
Ou seja, o efeito Aharonov-Bohm รฉ um caso particular da fase de Berry.
