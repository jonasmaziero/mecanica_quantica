# Problema de dois estados, <br> na representação de interação

Aqui apresentaremos exemplos de sistemas físicos para os quais o hamiltoniano não comuta em tempos diferentes, e para os quais, portanto, não podemos utilizar o operador de evolução temporal na forma $U_t = e^{-\frac{i}{\hbar}\int_0^t H_{t'}dt'}$. A princípio poderíamos tentar aplicar a série de Dyson nesses casos. Mas aqui aplicaremos a representação de interação (RI), a partir da qual conseguiremos obter as informações relevantes para os sistemas considerados. 

Vale observar também que aqui trataremos explicitamente a interação de um sistema quântico com um campo clássico. Isso nos possibilitará descrever não só a quantização dos níveis de energia, mas também a transição entre esses níveis de energia. Ou seja, aqui veremos aqui como um dos postulados de Bohr (das transições entre estados estácionários) aparece diretamente da aplicação dos postulados da Mecânica Quântica.

Devo chamar a atenção também para o fato de que os resultados obtidos aqui não se aplicam somente a sistemas de dois níveis, mas sim a qualquer sistema com $n$ níveis quando focamos em transições possíveis entre 2 desses $n$ níveis. 

Consideremos o hamiltoniano dependente do tempo
\begin{equation}
H_{t}=H_{c}+V_{t},
\end{equation}
onde
\begin{align}
H_{c} & = E_{1}^{c} |E_{1}^{c}\rangle\langle E_{1}^{c}| + E_{2}^{c} |E_{2}^{c}\rangle\langle E_{2}^{c}| =: e_{1} |1\rangle\langle 1| + e_{2} |2\rangle\langle 2| \\
& = \begin{bmatrix} e_{1} & 0 \\ 0 & e_{2} \end{bmatrix},
\end{align}
é a parte independente do tempo do Hamiltoniano (assumimos que a diagonalização de $H_{c}$ é conhecida) e
\begin{align}
V_{t} & = \gamma\left(e^{i\omega t} |1\rangle\langle 2| + e^{-i\omega t} |2\rangle\langle 1| \right) \\
& = \begin{bmatrix} 0 & \gamma e^{i\omega t} \\ \gamma e^{-i\omega t} & 0 \end{bmatrix},
\end{align}
é a parte dependente do tempo do Hamiltoniano, geralmente de interação com uma __campo clássico__ dependente do tempo.
Acima $e_{2}>e_{1}$ e as representações são feitas na __base padrão__ ou base computacional: $\{|j\rangle\}_{j=1}^{2}$. 

Note que $\gamma$ dita a _intensidade da interação_ e a frequência de algumas mudanças temporais que ocorrerão neste sistema. Vale observar também que
\begin{equation}
H_{t} = \begin{bmatrix} e_{1} & \gamma e^{i\omega t} \\ \gamma e^{-i\omega t} & e_{2} \end{bmatrix}
\end{equation}
é a matriz 2x2 Hermitiana mais geral que podemos construir. Em relação ao tempo, o que poderíamos fazer de mais geral é usar $e_{j}(t)$, $\gamma(t)$ e $\omega(t)$.

Pode-se verificar que esse hamiltoniano não comuta em tempos diferentes, ou seja, em geral
$$[H_{t},H_{t'}]\ne 0.$$
Mais especificamente, pode-se verificar que $[H_{t},H_{t'}]=0$ se e somente se $t=t'+n2\pi/\omega$ com $n\in\mathbb{N}.$ Por conseguinte, não poderemos utilizar $U_{t} = e^{-\frac{i}{\hbar}\int_{0}^{t}H(t')dt'}$. Como comentamos acima, esse sistema é um exemplo de situação onde é útil usar a representação de interação.

Seguindo, lembramos que, na RI, escrevemos
\begin{equation}
|\psi_{t}^{I}\rangle = \sum_{j=1}^{2} c_{j}(t)|{j}\rangle=c_{1}(t)|{1}\rangle+c_{2}(t)|{2}\rangle
\end{equation}
e chegamos na equação matricial para os coeficientes
\begin{align}
i\hbar\begin{bmatrix}\partial_{t}c_{1} \\ \partial_{t}c_{2}\end{bmatrix} & = \begin{bmatrix} V_{11} & V_{12}e^{i\omega_{12}t} \\ V_{21}e^{i\omega_{21}t} & V_{22} \\ \end{bmatrix}\begin{bmatrix}c_{1} \\ c_{2}\end{bmatrix}  = \begin{bmatrix} 0 & \gamma e^{i\omega t}e^{-i\omega_{21}t} \\ \gamma e^{-i\omega t}e^{i\omega_{21}t} & 0 \\ \end{bmatrix}\begin{bmatrix}c_{1} \\ c_{2}\end{bmatrix} \\
& = \begin{bmatrix}\gamma e^{i(\omega-\omega_{21}) t}c_{2} \\ \gamma e^{-i(\omega-\omega_{21}) t}c_{1}\end{bmatrix}
,
\end{align}
onde para o presente Hamiltoniano usamos que $\omega_{21} = -\omega_{12} = (e_{2}-e_{1})/\hbar$ e $V_{11} = V_{22} = 0$ e $V_{12} = V_{21}^{*} = \gamma e^{i\omega t}$.

Para desacoplar esse sistema de equações lineares, vamos tomar a derivada temporal dessas equações:
\begin{cases}
i\hbar\partial_{tt}c_{1} = i\gamma(\omega-\omega_{21})e^{i(\omega-\omega_{21})t}c_{2} + \gamma e^{i(\omega-\omega_{21})t}\frac{i\hbar}{i\hbar}\partial_{t}c_{2}, \\
i\hbar\partial_{tt}c_{2} = -i\gamma(\omega-\omega_{21})e^{-i(\omega-\omega_{21})t}c_{1} + \gamma e^{-i(\omega-\omega_{21})t}\frac{i\hbar}{i\hbar}\partial_{t}c_{1}.
\end{cases}
Usando as equações diferenciais de primeira ordem obteremos:
\begin{cases}
i\hbar\partial_{tt}c_{1} = i(\omega-\omega_{21})i\hbar\partial_{t}c_{1} + \gamma e^{i(\omega-\omega_{21})t}\frac{1}{i\hbar}\gamma e^{-i(\omega-\omega_{21}) t}c_{1}, \\
i\hbar\partial_{tt}c_{2} = -i(\omega-\omega_{21})i\hbar\partial_{t}c_{2} + \gamma e^{-i(\omega-\omega_{21})t}\frac{1}{i\hbar}\gamma e^{i(\omega-\omega_{21}) t}c_{2}.
\end{cases}
Rearranjamos essas equações para obter o seguinte conjunto de _equações diferenciais de segunda ordem mas desacopladas_:
\begin{cases}
\partial_{tt}c_{1} - i(\omega-\omega_{21})\partial_{t}c_{1} + \frac{\gamma^{2}}{\hbar^{2}}c_{1} = 0, \\
\partial_{tt}c_{2} + i(\omega-\omega_{21})\partial_{t}c_{2} + \frac{\gamma^{2}}{\hbar^{2}}c_{2} = 0.
\end{cases}
A seguir obtemos a solução geral desse tipo de equação diferencial.

---
Solução de equações diferenciais do tipo
\begin{equation}
\partial_{tt}y+b\partial_{t}y+cy=0
\end{equation} 
para $y=y(t)$ e $b,c$ constantes. Se acharmos duas soluções particulares $y_{1}$ e $y_{2}$ linearmente independentes, podemos construir uma solução geral $y=C_{1}y_{1}+C_{2}y_{2}$ com $C_{1},C_{2}\in\mathbb{C}$. Vamos testar uma solução do tipo
\begin{equation}
y = e^{rt} \therefore \partial_{t}y=re^{rt}=ry \therefore \partial_{tt}y=r^{2}e^{rt}=r^{2}y.
\end{equation}
Substituindo na equação diferencial vem que
\begin{equation}
r^{2}y+bry+cy=(r^{2}+br+c)y=0\therefore r = \frac{-b\pm\sqrt{b^{2}-4c}}{2} =: r_{\pm}.
\end{equation}
Temos com isso que $e^{r_{+}t}$ e $e^{r_{-}t}$ são soluções da equação diferencial acima. Se $r_{+}\ne r_{-}$ podemos construir uma solução geral como
\begin{equation}
y(t) = C_{1}e^{r_{+}t}+C_{2}e^{r_{-}t} = e^{-bt/2}\left(C_{1}e^{\sqrt{b^{2}-4c}t/2}+C_{2}e^{-\sqrt{b^{2}-4c}t/2}\right).
\end{equation}

Nos casos em que $r_{+}=r_{-}$, a solução geral é escrita como
\begin{equation}
y(t) = C_{1}e^{r_{+}t}+C_{2}te^{r_{-}t} = e^{-bt/2}\left(C_{1}e^{\sqrt{b^{2}-4c}t/2}+C_{2}te^{-\sqrt{b^{2}-4c}t/2}\right).
\end{equation}

Abaixo mostramos como o _Sympy_ pode ser utilizado para obter a _solução_ desse tipo de equação diferencial.

```python
from sympy import symbols, Function, dsolve, Derivative, diff, init_printing
init_printing(use_unicode=True)
t, w, g, w21, hb = symbols('t omega gamma \\omega_{21} hbar')
t, w, g, w21, hb
```

```python
c1 = Function('c1')
dsolve(Derivative(c1(t),t,t) - 1j*(w-w21)*Derivative(c1(t),t)\
       + (g**2/hb**2)*c1(t), c1(t))
```

```python
c2 = Function('c2')
dsolve(diff(c2(t),t,t) + 1j*(w-w21)*diff(c2(t),t) + (g**2/hb**2)*c2(t), c2(t))
```

---

Utilizando a solução geral acima, e aplicando para a equação diferencial para $c_{1}$, temos 
$$b=-i(\omega-\omega_{21}),\ c=\gamma^{2}/\hbar^{2}.$$
Assim
\begin{align}
c_{1}(t) & = e^{i(\omega-\omega_{21})t/2}\left(C_{1}e^{\sqrt{-(\omega-\omega_{21})^{2}-4\gamma^{2}/\hbar^{2}}t/2}+C_{2}e^{-\sqrt{-(\omega-\omega_{21})^{2}-4\gamma^{2}/\hbar^{2}}t/2}\right) \\
& = e^{i(\omega-\omega_{21})t/2}\left(C_{1}e^{i\sqrt{((\omega-\omega_{21})/2)^{2}+\gamma^{2}/\hbar^{2}}t}+C_{2}e^{-i\sqrt{((\omega-\omega_{21})/2)^{2}+\gamma^{2}/\hbar^{2}}t}\right).
\end{align}
Da equação diferencial para $c_{2}$, temos 
$$b=i(\omega-\omega_{21}),\ c=\gamma^{2}/\hbar^{2}.$$
Então
\begin{equation}
c_{2}(t) = e^{-i(\omega-\omega_{21})t/2}\left(C'_{1}e^{i\sqrt{((\omega-\omega_{21})/2)^{2}+\gamma^{2}/\hbar^{2}}t}+C'_{2}e^{-i\sqrt{((\omega-\omega_{21})/2)^{2}+\gamma^{2}/\hbar^{2}}t}\right).
\end{equation}

Definindo
\begin{equation}
\xi := \sqrt{\frac{\gamma^{2}}{\hbar^{2}}+\left(\frac{\omega-\omega_{21}}{2}\right)^{2}}. 
\end{equation}
reescrevemos as soluções acima como
\begin{align}
c_{1}(t) & = e^{i(\omega-\omega_{21})t/2}\left(C_{1} e^{-i\xi t} + C_{2}e^{i\xi t}\right), \\
         & = e^{i(\omega-\omega_{21})t/2}\left[(C_{1}+C_{2})\cos\xi t + i(C_{2}-C_{1})\sin\xi t\right], \\
         & =: e^{i(\omega-\omega_{21})t/2}\left(k_{1}\sin\xi t + k_{2}\cos\xi t \right), \\
c_{2}(t) & = e^{-i(\omega-\omega_{21})t/2}\left(k_{3}\sin\xi t + k_{4}\cos\xi t \right),
\end{align}
As constantes $\{k_{j}\}_{j=1}^{4}$ devem ser determinadas impondo as condições física sobre a função de onda. Neste caso usaremos as condições iniciais e de normalização.

__Especificando o estado inicial__ <br>
Vamos começar assumindo que o sistema está _inicialmente_ no estado fundamental do Hamiltoniano independente do tempo $H_{c}$, i.e.,
\begin{equation}
|\psi_{0}^{I}\rangle = |\psi^{s}_{0}\rangle = c_1(0)|1\rangle + c_2(0)|2\rangle = |1\rangle.
\end{equation}
Nesse caso, devemos ter
\begin{align}
c_{1}(0) & = e^{0}\left(k_{1}\sin(0) + k_{2}\cos(0)\right) = k_{2} = 1, \\
c_{2}(0) & = e^{0}\left(k_{3}\sin(0) + k_{4}\cos(0)\right) = k_{4} = 0.
\end{align}
Vem assim que
\begin{equation}
c_{2}(t) = e^{-i(\omega-\omega_{21})t/2}k_{3}\sin\xi t.
\end{equation}
Para facilitar a solução desse problema, usamos uma das equações do sistema original, $i\hbar\partial_{t}c_{2}(t)= \gamma e^{-i(\omega-\omega_{21}) t}c_{1}(t)$, para escrever
\begin{align}
c_{1}(t) & = \gamma^{-1}e^{i(\omega-\omega_{21}) t}i\hbar\partial_{t}c_{2}(t) \\
         & = \gamma^{-1}e^{i(\omega-\omega_{21}) t}i\hbar\left(-i\frac{\omega-\omega_{21}}{2}e^{-i(\omega-\omega_{21})t/2}k_{3}\sin\xi t  + \xi e^{-i(\omega-\omega_{21})t/2}k_{3}\cos\xi t\right) \\
         & = \gamma^{-1}i\hbar k_{3}e^{i(\omega-\omega_{21}) t/2}\left(-i\frac{\omega-\omega_{21}}{2}\sin\xi t  + \xi\cos\xi t\right).
\end{align}

Agora aplicamos a normalização do estado $|\psi_{t}^{I}\rangle=c_{1}|1\rangle+c_{2}|2\rangle$:
\begin{align}
& |c_{1}(t)|^{2} + |c_{2}(t)|^{2} = 1 \\
& \therefore\ \gamma^{-2}\hbar^{2} |k_{3}|^{2}\left(-i\frac{\omega-\omega_{21}}{2}\sin\xi t  + \xi\cos\xi t\right)\left(i\frac{\omega-\omega_{21}}{2}\sin\xi t  + \xi\cos\xi t\right) + |k_{3}|^{2}\sin^{2}\xi t = 1 \\
& \therefore\ \gamma^{-2}\hbar^{2} |k_{3}|^{2}\left(\left(\frac{\omega-\omega_{21}}{2}\right)^{2}\sin^{2}\xi t  + \xi^{2}\cos^{2}\xi t\right) + |k_{3}|^{2}\sin^{2}\xi t = 1 \\ 
& \therefore\ |k_{3}|^{2}\left(\left(\left(\frac{\omega-\omega_{21}}{2}\right)^{2}\sin^{2}\xi t  + \xi^{2}\cos^{2}\xi t\right) + (\gamma^{2}/\hbar^{2})\sin^{2}\xi t \right)= \gamma^{2}/\hbar^{2}.
\end{align}

Assim
\begin{align}
|k_{3}|^{2} & = \frac{\gamma^{2}/\hbar^{2}}{\left(\left(\frac{\omega-\omega_{21}}{2}\right)^{2}+\frac{\gamma^{2}}{\hbar^{2}}\right)\sin^{2}\xi t  + \xi^{2}\cos^{2}\xi t} \\
& = \frac{\gamma^{2}/\hbar^{2}}{\xi^{2}\sin^{2}\xi t  + \xi^{2}\cos^{2}\xi t} \\
& = \frac{\gamma^{2}/\hbar^{2}}{\gamma^{2}/\hbar^{2} + \left(\frac{\omega-\omega_{21}}{2}\right)^{2}} \\
& = \frac{1}{1 + \left(\frac{\hbar\omega_{21}}{2\gamma}\left(1-\frac{\omega}{\omega_{21}}\right)\right)^{2}}.
\end{align}

A __probabilidade de transição__ $|1\rangle\rightarrow|2\rangle$, que é igual a probabilidade de, em uma medida de energia no instante de tempo $t$, obtermos $e_{2}$ é:
\begin{align}
Pr(2(t)) & = \left|\langle 2|\psi_{t}^{I}\rangle\right|^{2} \\
& = \left|c_{2}(t)\right|^{2} \\
& = |k_{3}|^{2}\sin^{2}\xi t \\
& = \frac{\sin^{2}\xi t}{1 + \left(\frac{\hbar\omega_{21}}{2\gamma}\left(1-\frac{\omega}{\omega_{21}}\right)\right)^{2}}.
\end{align}
Notamos que as _probabilidades oscilam sinudoisalmente_ com o tempo, com amplitude máxima determinada por $k_{3}$. 

```python
from matplotlib import pyplot as plt
from ipywidgets import interactive
import numpy as np
import math

def pr2t(g,w):
    hbb = 1; w21 = 1; t = np.arange(0,10,0.05)
    k32 = 1/(1+(((hbb*w21)/(2*g))*(1-w/w21))**2)
    csi = math.sqrt((g/hbb)**2+((w-w21)/2)**2)
    pr2 = k32*(np.sin(csi*t)**2)
    plt.plot(t, pr2); plt.xlabel(r'$t$')
    plt.ylabel(r'$Pr(2(t))$')
    plt.annotate(r'$\gamma=$'+str(g)[:4], xy=(8,0.9), \
                 xytext=(8,0.9), fontsize=13)
    plt.annotate(r'$\omega=$'+str(w)[:4], xy=(8,0.8), \
                 xytext=(8,0.8), fontsize=13)
    plt.xlim([0,10]); plt.ylim([0,1]); plt.show()

interactive(pr2t, g = (0.01,3,0.1), w = (0,5,0.1))
```

Ademais, só teremos probabilidade máxima de encontrar o sistema no estado excitado se $\omega=\omega_{21}$ (que resulta em $k_{3}=1$), que é a __condição de ressonância__. Ou seja, o termo dependente do tempo do Hamiltoniano terá eficiência máxima em causar transição entre os níveis de energia quando sua frequência for igual à frequência de Bohr do sistema, ou seja, quando estivermos na chamada condição de ressonância. 

A dependência da probabilidade máxima
\begin{equation}
\max_{t}Pr(2(t)) = |k_{3}|^{2}
\end{equation}
com a razão entre as escalas de intensidade das interações de $H_{c}$ e de $V_{t}$, i.e., $\hbar\omega_{21}/\gamma$ e entre as frequências "aplicada" e natural do sistema pode ser estudada com os gráficos produzidos com o código abaixo. Os tempos para os quais a probabilidade é máxima são:
\begin{equation}
\xi t_{\max}=(2j+1)\frac{\pi}{2} \therefore t_{\max} = \frac{(2j+1)\pi}{2\xi} \text{ com } j=0,1,2,3,\cdots.
\end{equation}

Note que esse resultado indica que se a intensidade de acoplamento $\gamma$ é muito pequena, quase que somente a frequência de ressonância $\omega=\omega_{21}$ será capaz de causar transição de estados. Quanto maior for $\gamma$ mais valores de $\omega$ pederão levar a transições $|1\rangle\leftrightarrow |2\rangle$.

```python
def pr2m(g):
    hbb = 1; w21 = 20; w = np.arange(0,40,0.01)
    k32 = 1/(1+(((hbb*w21)/(2*g))*(1-w/w21))**2)
    plt.plot(w, k32); plt.xlabel(r'$\omega/\omega_{21}$')
    plt.ylabel(r'$Pr(2(t))_{\max}$')
    plt.annotate(r'$\gamma=$'+str(g)[:4], xy=(30,0.8), \
                 xytext=(30,0.8), fontsize=13)
    plt.xlim([0,40]); plt.ylim([0,1.01]); plt.show()

interactive(pr2m, g = (0.01,2,0.05))
```

## Ressonância magnética de spin
Consideremos uma partícula de spin 1/2 sujeita ao seguinte campo magnético
\begin{equation}
\vec{B}(t) = B_{0}\hat{k} - B_{1}\left(\hat{i}\cos\omega t - \hat{j}\sin\omega t\right),
\end{equation}
com $B_{0}$ e $B_{1}$ constantes positivas. Note que há um termo constante na direção $z$ e outro termo que gira no plano $xy$. Para esta interação o Hamiltoniano será
\begin{align}
H_{t} & = -\vec{\mu}\cdot\vec{B}(t) \\
& = -\frac{\hbar gq}{4m}\vec{\sigma}\cdot\vec{B}(t) \\
& = -\frac{\hbar gq}{4m}(\sigma_{x}\hat{i}+\sigma_{y}\hat{j}+\sigma_{z}\hat{k})\cdot\left(B_{0}\hat{k} - B_{1}\left(\hat{i}\cos\omega t - \hat{j}\sin\omega t\right)\right) \\
& = -\frac{\hbar gq B_{0}}{4m}\sigma_{z} + \frac{\hbar gq B_{1}}{4m}(\sigma_{x}\cos\omega t - \sigma_{y}\sin\omega t) \\
& = -\frac{\hbar\omega_{21}}{2}\sigma_{z} + \gamma(\sigma_{x}\cos\omega t - \sigma_{y}\sin\omega t) \\
& = e_{1}|z_{+}\rangle\langle z_{+}| + e_{2}|z_{-}\rangle\langle z_{-}| + \gamma[(|z_{+}\rangle\langle z_{-}|+|z_{-}\rangle\langle z_{+}|)\cos\omega t - (-i|z_{+}\rangle\langle z_{-}|+i|z_{-}\rangle\langle z_{+}|)\sin\omega t)] \\
  & = H_{c} + \gamma[|z_{+}\rangle\langle z_{-}|(\cos\omega t +i\sin\omega t) + |z_{-}\rangle\langle z_{+}|(\cos\omega t -i\sin\omega t)] \\
  & = H_{c} + \gamma(|z_{+}\rangle\langle z_{-}|e^{i\omega t} + |z_{-}\rangle\langle z_{+}|e^{-i\omega t}) \\
  & = H_{c} + V_{t},
\end{align}
onde definimos 
$$\omega_{21} = \frac{gqB_{0}}{2m} \text{ e } \gamma = \frac{\hbar qgB_{1}}{4m}.$$

Como esse Hamiltoniano é idêntico ao que consideramos na última seção, todos os resultados obtidos lá se aplicam aqui. Sabemos que se $B_{1}=0$ o spin precessará em torno da direção $z$. No entando, se $B_{1}>0$ teremos "flips" (__transições__) entre os estados
\begin{equation}
|z_{+}\rangle\leftrightarrow|z_{-}\rangle.
\end{equation}
A freqûencia com que essas transições ocorrerão depende das razões $\gamma/\omega_{21}$ e $\omega/\omega_{21}$, como discutido acima.

A ressonância magnética nuclear (ou qualquer outro tipo de ressonância magnética) consiste de, dentro do contexto dos cálculos que fizemos, aplicar $V_{t}$ (ligar $B_1$) por um tempo $\tau$ suficiente para que $c_{1}=c_{2}=1/\sqrt{2}$, e assim preparar o quanton no estado de momentos magnético:
\begin{equation}
\frac{1}{\sqrt{2}}|z_{+}\rangle + \frac{1}{\sqrt{2}}|z_{-}\rangle \equiv |x_{+}\rangle.
\end{equation}
Depois, para um ensemble com muitos spins, mede-se momento magnético médio (através do campo magnético gerado) em alguma direção no plano $xy$. Como $B_{1}(t>\tau)=0$, os spins simplesmente precessariam em torno da direção $z$, 
\begin{align}
\langle\sigma_{x}(t)\rangle & = \langle\sigma_{x}\rangle\cos(\omega_{21} t) + \langle\sigma_{y}\rangle\sin(\omega_{21} t) \\
& = \langle\sigma_{x}\rangle\cos(\omega_{21} t) + \langle\sigma_{y}\rangle\cos(\omega_{21} t-\pi/2),
\end{align}
como mostrado na figura abaixo.

```python
t = np.arange(0,10*math.pi,0.1); w21 = 1
r1 = 1/math.sqrt(2); r2 = r1
r1t = r1*np.cos(w21*t) - r2*np.sin(w21*t); plt.plot(t,r1t)
plt.xlabel(r'$t$'); plt.ylabel(r'$\langle\sigma_{x}\rangle$')
plt.show()
```

No entanto, a temperatura é diferente de zero e todo sistema deixado evoluir livremente sob a ação de um banho térmico vai trocar energia com este último e estará, depois de um tempo longo, em um dos auto-estados do Hamiltoniano, $|z_{\pm}\rangle$, com probabilidade $\propto e^{-\beta E_{\pm}}$ com $\beta=1/k_{B}T$, onde $k_{B}$ é a constante de Boltzmann e $T$ é a temperatura do ambiente. Conforme o tempo passa, e mais e mais spins passam para os estados $|z_{\pm}\rangle$, o momento magnético médio diminui. Esse decaimento é exponencial. Exemplificamos esse comportamento,
\begin{align}
\langle\sigma_{x}(t)\rangle = \big(\langle\sigma_{x}\rangle\cos(\omega_{21} t) + \langle\sigma_{y}\rangle\cos(\omega_{21} t-\pi/2)\big)e^{-\Gamma_{1}t},
\end{align}
com a figura abaixo.

```python
t = np.arange(0,100*math.pi,0.1)
T1 = 100; a = (1-np.exp(t/T1))/2
w1 = 1; w2 = 10; r1 = 1/math.sqrt(2); r2 = r1
r1t = (r1*np.cos(w1*t) - r2*np.sin(w1*t))*np.exp(-t/T1) \
+ (r1*np.cos(w2*t) - r2*np.sin(w2*t))*np.exp(-t/T1)
plt.xlabel(r'$t$');plt.ylabel(r'$\langle\sigma_{x}\rangle$')
plt.plot(t,r1t); plt.show()
```

A frequência de precessão do spin é proporcional ao campo magnético na direção $z$. Em uma molécula, dependendo do ambiente químico que um certo tipo de átomo tem em uma molécula, diferentes átomos desse tipo poderão estar sujeitos a diferentes campos magnéticos, que correspondem a frequências diferentes. Essas frequências são identificadas tomando-se a transformada de Fourier de $\langle\sigma_{x}(t)\rangle$. E é assim que podemos identificar a presença de diferentes átomos, ou de diferentes grupos de átomos, em uma certa amostra.

```python
from scipy import fftpack
ft = fftpack.fft(r1t); plt.plot(t,ft.real)
plt.xlabel(r'$\omega$'); plt.show()
```

## MASER
Alguns sistemas, como e.g. a __molécula de amônia__ $NH_{3}$, possuem momento de dipolo elétrico, que, no regime quântico, interage com campos elétricos segundo o Hamiltoniano (veja a Ref. 4 para mais detalhes):
\begin{equation}
H = -\vec{\mu}_{e}\cdot\vec{E}.
\end{equation}
Se experimentos do tipo Stern-Gerlach são feitos com essas partículas (com gradientes de campo elétrico), os resultados são idênticos àqueles obtidos para partículas de spin 1/2. Portanto os observáveis e estados são descritos pela "mesma matemática". Por conseguinte, se aplicamos um campo elétrico
\begin{equation}
\vec{E}(t) = E_{0}\hat{k} - E_{1}\left(\hat{i}\cos\omega t - \hat{j}\sin\omega t\right)
\end{equation}
podemos obter transições entre os estados de momento de dipolo elétrico, tal qual o caso da ressonância magnética da última seção.

No caso da amônia $\mathrm{NH_{3}}$, a frequência de Bohr $\omega_{21}$ está no intervalo de microondas:
\begin{equation}
\nu=(300\text{ MHz} - 300\text{ GHz}) \text{ ou } \lambda = (100\text{ cm} - 0,1\text{ cm}).
\end{equation}
O procedimento para obter o MASER (__Microwave Amplification by Stimulated Emission of Radiation__) (https://www.nobelprize.org/prizes/physics/1964/summary/) é o seguinte (veja a figura abaixo, retirada da Ref. 4).

Primeiro faz-se as moléculas passarem por um gradiente de campo elétrico na direção $z$, "selecionando" aquelas que são desviadas pra baixo naquela direção:
\begin{equation}
|?\rangle\rightarrow \boxed{SG_{z}}\begin{matrix}\rightarrow|z_{+}\rangle \rightarrow \text{descarta} \\ \rightarrow|z_{-}\rangle \rightarrow \text{seleciona}\end{matrix}.
\end{equation}
Ou seja, preparamos todas as moléculas com componente de dipolo elétrico negativa na direção $z$, i.e., as preparamos no estado $|z_{-}\rangle\equiv|2\rangle$, que é o estado de maior energia.

Depois faz-se essas moléculas passarem por uma cavidade onde aplica-se um campo oscilante no plano $xy$ (radiação de microondas), como o descrito na equação acima, na condição de ressonância $\omega=\omega_{21}$. Pode-se verificar que se a cavidade é construída de forma que as moléculas passem um tempo
\begin{equation}
t_{opt} = \frac{\pi\hbar}{2\gamma},
\end{equation}
todas as moléculas terão decaído para o estado fundamental e emitido radiação de microondas antes de sair da cavidade. Isso fará com que a radiação de microondas na cavidade seja amplificada, ou seja, teremos mais luz com aquela frequência, e isso se deve ao estímulo do campo externo.

Para verificar que o tempo é dado como acima, notamos que 
$$k_{1}(\omega = \omega_{21}) = 1$$
e portanto 
$$Pr(1(t)) = k_{1}^{2}\sin^{2}\xi t = \sin^{2}(\gamma t/\hbar),$$
que é máxima quando 
$$\frac{\gamma t}{\hbar} = \frac{(2j+1)\pi}{2} \text{ para } j=0,1,2,3,\dots.$$
Para $j=0$ obtemos a expressão acima, para o menor tempo necessário para que $Pr(1(t))=1$.

### Teoria de Einstein da radiação
Consideremos um _gás de átomos_ na presença de radiação eletromagnética (REM), dentro de uma cavidade, por exemplo. Consideremos dois níveis de energia $E_{m}$ e $E_{n}$ de átomos, com 
$$E_{m}>E_{n}.$$
Segundo a teoria de Bohr, um átomo sujeito a uma REM poderá absorver um fóton de frequência
\begin{equation}
\nu_{mn}=\frac{E_{m}-E_{n}}{h},
\end{equation}
transitando assim do estado de menor energia $|E_{n}\rangle$ para o estado de maior energia $|E_{m}\rangle$. Se a intensidade da REM aplicada é $I(\nu)$, na frequência $\nu$, e $N_{n}$ é o número de átomos no estado $|E_{n}\rangle$, em um certo instante de tempo, a taxa temporal com que essas transições acontecem (número de transições $|E_{n}\rangle\rightarrow|E_{m}\rangle$ por segundo) pode ser escrita como
\begin{equation}
\partial_{t}N_{n} \propto N_{n}I(\nu) = B_{nm}N_{n}I(\nu).
\end{equation}

A REM também induz o átomo a fazer a transição de estados de maior energia para estados de menor energia, $|E_{m}\rangle\rightarrow|E_{n}\rangle$, em um processo denominado __emissão estimulada__. Nesse caso o átomo emite um fóton de frequência $\nu_{mn}=(E_{m}-E_{n})/h$.
Se, em um certo instante de tempo $t$, temos $N_{m}$ átomos no estado de maior energia, a taxa com que estes átomos decaem por emissão estimulada é
\begin{equation}
\partial_{t}N_{m}\propto N_{m}I(\nu) = B_{mn}N_{m}I(\nu).
\end{equation}

$B_{mn}$ e $B_{nm}$ são os chamados __coeficientes $B$ de Einstein__. Einstein usou conservação do momento linear para argumentar que 
$$B_{mn}=B_{nm}$$
(veja a tradução de (Physikalische Zeitschrift18, 121 (1917)) em (Rev. Bras. Ens. Fis. 27, 93 (2005))). Na realidade, nas condições que consideramos aqui, e usando a suposição de Einstein de que 
$$N_{n}(t)=N_{m}(t),$$
verificamos explicitamente essa previsão e identificamos esses coeficientes:
\begin{align}
B_{nm}N_{n}I(\nu) & = \partial_{t}N_{n} = Pr(|1\rangle\rightarrow |2\rangle) = k_{3}^{2}\sin^{2}(\xi t), \\
B_{mn}N_{m}I(\nu) & = \partial_{t}N_{m} = Pr(|2\rangle\rightarrow |1\rangle) = k_{1}^{2}\sin^{2}(\xi t), \\
\end{align}
com
\begin{equation}
|k_{1}|^{2}=|k_{3}|^{2} = \left(1+\left(\frac{\hbar\omega_{21}}{2\gamma}\left(1-\frac{\omega}{\omega_{21}}\right)\right)^{2}\right)^{-1}.
\end{equation}
Acima, para obter as igualdades, assumimos que as populações atômicas e a intensidade da REM foram normalizadas adequadamente.

Além destes dois processos identificados, para trocas de energia entre radiação eletromagnética e matéria, existe outro efeito importante que é a __emissão espontânea__, que é explicada quantizando tanto a matéria como a radiação eletromagnética, em uma teoria denominada __Óptica Quântica__.
