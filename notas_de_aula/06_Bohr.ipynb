A __teoria eletromagnética clássica__, que associa a frequência da luz emitida por um material com a _frequência de oscilação das cargas elétricas_ que o compõe, não consegue de forma alguma explicar a existência de espectros de linhas (que seria equivalente a explicar por quê as cargas elétricas só oscilam em um conjunto discreto de frequências). A explicação para esse fenômeno é obtida entendendo-se melhor a estrutura do átomo, como veremos a seguir.

## A estrutura do átomo
Sabendo que os átomos continham elétrons, cuja carga elétrica é $-e$ e a massa $m_{e}$ é muito menor que a massa do átomo, __J.J. Thomson__ propôs o _modelo atômico do pudim de passas_, em que a carga positiva massiva é o pudim e os elétrons são as passas, como ilustrado na figura abaixo.
<figure>
    <img src="fig_04_thomson_pudim.png" width="200">
    <div class="caption">Fonte: https://commons.wikimedia.org/wiki/File:PlumPuddingModel_ManyCorpuscles.png.</div>
</figure>
Existem vários problemas com esse modelo. Segundo a teoria eletromagnética clássica, não existe configuração de equilíbrio estável para cargas elétricas distribuídas dessa forma. Além disso, Thomson não conseguiu encontrar os modos normais de vibração das cargas elétricas cujas frequências corresponderiam e.g. ao espectro de linhas do átomo de Hidrogênio.

Um passo importante para o entendimento da estrutura atômica foi dado por __Rutherford__, que observou experimentalmente que partículas $\alpha\equiv He^{++}$ (átomos de Hélio duplamente ionizados) altamente energéticas podiam ser refletidas de $180^{\circ}$ quando incidiam em uma folha de ouro com aproximadamente $2000$ átomos de espessura. A única explicação para tal resultado é que a massa do átomo deve estar concentrada em uma região muito pequena, seu núcleo, com carga positiva, e que os elétrons, muito leves, orbitam esse núcleo, como ilustrado na figura abaixo.
<figure>
    <img src="fig_04_rutherford_atom.png" width="200">
    <div class="caption">Fonte: Adaptado de https://commons.wikimedia.org/wiki/File:Rutherford_atom.svg.</div>
</figure>

No entanto, esse modelo também tem problemas pois, segundo a teoria eletromagnética clássica, o elétron, que está acelerado, deveria emitir radiação eletromagnética, perdendo energia continuamente e por fim colapsando no núcleo. A seguir veremos como Bohr contornou esse problema.

## Modelo de Bohr do átomo
Bohr considerou que o único elétron de um átomo se movimenta em torno de um núcleo com $Z$ prótons ($Z$ é o número atômico do átomo) (https://www.nobelprize.org/prizes/physics/1922/bohr/lecture/). Além disso, ele usou axiomas equivalentes aos seguintes postulados:
1. O elétron se move em trajetórias circulares sem irradiar energia, nos chamados __estados estacionários__, com momento angular quantizado:
\begin{equation}
L = L_{n} = n\hbar
\end{equation}
com 
$$n=1,2,3,\cdots$$
e
\begin{equation}
\hbar:=\frac{h}{2\pi}.
\end{equation}
1. O átomo irradia ou absorve energia somente quando o elétron passa de um estado estacionário para outro. A frequência do fóton é obtida por conservação de energia:
\begin{equation}
h\nu=|E_{i}-E_{f}|,
\end{equation}
com $E_{i}$ e $E_{f}$ sendo as energias dos estados estacionários inicial e final, respectivamente.

O modelo atômico de Bohr está ilustrado na figura abaixo.
<figure>
    <img src="fig_04_bohr_atom.png" width="250">
    <div class="caption">Fonte: https://commons.wikimedia.org/wiki/File:Bohr_atom_model.svg.</div>
</figure>

O elétron é mantido na órbita circular pela força centrípeda dada pela força elétrica:
\begin{equation}
|\vec{F}_{e}|=|\vec{F}_{c}|\ \therefore\ \frac{k(Ze)e}{r^{2}}=\frac{m_{e}v^{2}}{r},
\end{equation}
com
\begin{equation}
k=8,9875517923\mathrm{x}10^{9} \text{ Nm}^{2}\text{/C}^{2}
\end{equation}
sendo a constante elétrica de Coulomb. Usando 
$$v=\sqrt{\frac{kZe^{2}}{m_{e}r}}$$
obtido da última equação e o postulado 1, podemos escrerver:
\begin{equation}
L=rp\sin(\pi/2)=rm_{e}v=rm_{e}\sqrt{\frac{Zke^{2}}{m_{e}r}}=\sqrt{m_{e}rZke^{2}}=n\hbar.
\end{equation}
Vemos assim que os raios, quantizados, das __órbitas de Bohr__ são dados por:
\begin{equation}
r = r_{n} = \left(\frac{\hbar^{2}}{m_{e}ke^{2}}\right)\frac{n^{2}}{Z}=:\frac{a_{0}}{Z}n^{2},
\end{equation}
com o __raio de Bohr__ definido como
\begin{equation}
a_{0} := \frac{\hbar^{2}}{m_{e}ke^{2}} \approx 0,529177\ \overset{\circ}{A}.
\end{equation}

```python
h = 6.626069*10**-34 # Js (constante de Planck)
import math
hb = h/(2*math.pi) # hbar
me = 9.109382*10**-31 # kg (massa do elétron)
k = 8.9875517923*10**9 # Nm^2/C^2 (constante de Coulomb)
e = 1.60217662*10**-19 # C (carga elementar)
```

```python
a0=(hb**2)/(me*k*e**2); print(a0)
```

A __energia__ total do átomo (consideramos o núcleo como estando em repouso) será então:
\begin{align}
E & = E_{c}+E_{p} \\
& = \frac{m_{e}v^{2}}{2} - \frac{kZe^{2}}{r} \\
& = \frac{k(Ze)e}{2r} - \frac{kZe^{2}}{r} \\ 
& = -\frac{k(Ze)e}{2r} \\
& = -\frac{k(Ze)e}{2}\frac{Z}{a_{0}n^{2}} \\
& = -\frac{kZ^{2}e^{2}}{2a_{0}}\frac{1}{n^{2}}.
\end{align}
Ou 
\begin{equation}
\color{blue}{E_{n} = \frac{Z^{2}E_{1}}{n^{2}}},
\end{equation}
com a energia do estado fundamental para $n=1$ sendo:
\begin{equation}
E_{1} := -\frac{ke^{2}}{2a_{0}} \approx -13,61\text{ eV}.
\end{equation}
Abaixo está mostrado o diagrama de níveis de energia e parte do espectro de linhas do átomo de Hidrogênio segundo o modelo de Bohr. O acordo é muito bom entre a teoria e o espectro de linhas mencionado acima.

```python
def bohr(Z,n): # em eV
    return ((-((k*e**2)/(2*a0))/e)*Z**2)/n**2
```

```python
Z=1; (bohr(Z,1),bohr(Z,2),bohr(Z,3),bohr(Z,4),bohr(Z,5))
```

```python
def bohr_ed(Z,xmin,xmax):
    import matplotlib
    matplotlib.rcParams.update({'font.size':12}); 
    plt.figure(figsize = (6,4), dpi = 100)
    plt.hlines(bohr(Z,1),xmin,xmax,color='magenta')
    plt.text(-0.15,bohr(Z,1),'1',color='magenta'); 
    plt.text(1.01,bohr(Z,1),'%3.2f' % (bohr(Z,1)),color='magenta')
    plt.hlines(bohr(Z,2),xmin,xmax,color='blue')
    plt.text(-0.15,bohr(Z,2),'2',color='blue'); 
    plt.text(1.01,bohr(Z,2),'%3.2f' % (bohr(Z,2)),color='blue')
    plt.hlines(bohr(Z,3),xmin,xmax,color='cyan') 
    plt.text(-0.15,bohr(Z,3),'3',color='cyan'); 
    plt.text(1.01,bohr(Z,3),'%3.2f' % (bohr(Z,3)),color='cyan')
    plt.hlines(bohr(Z,4),xmin,xmax,color='green')
    plt.text(-0.15,bohr(Z,4),'4',color='green'); 
    plt.text(1.01,bohr(Z,4),'%3.2f' % (bohr(Z,4)),color='green')
    plt.hlines(bohr(Z,5),xmin,xmax,color='gray'); 
    plt.hlines(bohr(Z,6),xmin,xmax,color='brown')
    plt.hlines(bohr(Z,7),xmin,xmax,color='brown'); 
    plt.hlines(bohr(Z,8),xmin,xmax,color='brown')
    plt.hlines(bohr(Z,9),xmin,xmax,color='gray'); 
    plt.hlines(bohr(Z,10),xmin,xmax,color='brown')
    plt.hlines(bohr(Z,11),xmin,xmax,color='brown'); 
    plt.hlines(0,0,1,color='red')
    plt.text(-0.15,0,r'$n=\infty$',color='red'); 
    plt.text(1.01,0.1,r'0 (eV)',color='red')
    plt.annotate("", xy=(0.5,bohr(Z,1)), xytext=(0.5,bohr(Z,2)), 
                 arrowprops=dict(arrowstyle="<->"))
    plt.text(0.51,(bohr(Z,1)+bohr(Z,2))/2,(c*heV)/(bohr(Z,2)-bohr(Z,1)))
    plt.annotate("", xy=(0.7,bohr(Z,2)), xytext=(0.7,bohr(Z,3)), 
                 arrowprops=dict(arrowstyle="<->"))
    plt.text(0.71,(bohr(Z,2)+bohr(Z,3))/2,(c*heV)/(bohr(Z,3)-bohr(Z,2)))
    plt.axis('off'); plt.xlim(0,2); plt.ylim(bohr(Z,1)-0.1,0); 
    plt.show()
```

```python
from matplotlib import pyplot as plt
c = 2.99792458*10**8 # m/s (velocidade da luz no vácuo)
heV = h/e # em eV

# Diagrama de níveis de energia no modelo de Bohr
# Z=1 => H, Z=2 => He^+, Z = 3 => Li^++ 
Z=1; bohr_ed(Z,0,1)
```

```python
def bohrf(Z,ni,nf): # Frequências de Bohr
    return abs((bohr(Z,nf)-bohr(Z,ni))/heV)
def bohrl(Z,ni,nf): # Comprimentos de onda de Bohr
    return c/bohrf(Z,ni,nf)
Z=1; (bohrl(Z,2,3),bohrl(Z,2,4),bohrl(Z,2,5),bohrl(Z,2,6),bohrl(Z,2,7),bohrl(Z,2,8),bohrl(Z,2,10**3))
```

## Experimento de Franck e Hertz
James Franck (https://www.nobelprize.org/prizes/physics/1925/franck/lecture/) e Gustav Hertz (sobrinho de Heinrich Hertz) (https://www.nobelprize.org/prizes/physics/1925/hertz/lecture/) fizeram um experimento decisivo para a aceitação de uma das consequências principais da teoria de Bohr do átomo, a quantização da energia. O aparato experimental e os resultados obtidos estão ilustrados na figura abaixo.

<figure>
    <img src="fig_04_franck-hertz.png" width="300">
    <div class="caption">Fonte: https://commons.wikimedia.org/wiki/File:Experiencia_Franck_Hertz.svg.</div>
</figure>

Nesse experimento, o filamento $H$ é usado para aquecer o cátodo $C$, arrancando assim elétrons dele. Esses elétrons são acelerados pela ddp $V_{0}$ aplicada entre $C$ e a grade $G$. Passando pela grade, os elétrons que possuem energia cinética maior que a barreira de potencial $e|V_{p}|$ chegam até a placa $P$, contribuindo assim para a corrente elétrica que é medida pelo amperímetro $A$. 

Se aumentamos o potencial de aceleração $V_{0}$, aumentamos a energia cinética dos elétrons que passam por $G$. Assim, se fazemos vácuo no tubo, se espera um aumento linear da corrente elétrica com $V_{0}$. No entanto, ao colocar um gás de mercúrio no tubo ($Hg$ possui 80 elétrons), observa-se o comportamento "oscilatório" da corrente elétrica, como mostrado na figura abaixo. 

<figure>
    <img src="fig_04_franck-hertz_.png" width="300">
    <div class="caption">Fonte: https://commons.wikimedia.org/wiki/File:Dades_Experiment_Franck_Hertz.png.</div>
</figure>

Podemos explicar esses resultados assumindo quantização da energia do átomo. Se o intervalo de energia entre o estado fundamental e primeiro estado excitado do $Hg$ é $4,9\text{ eV}$, então para $V_{0}<4,9\text{ V}$ o elétron incidente não possui energia suficiente para excitar elétrons do $Hg$. Para $V_{0}\ge 4,9\text{ V}$ os elétrons incidentes excitam os elétrons do $Hg$, perdendo $4,9\text{ eV}$ de sua energia, o que faz com que eles não tenham energia cinética suficiente para vencer a barreira de potencial $e|V_{p}|$, o que implica numa diminuição da corrente elétrica. Os outros picos de corrente elétrica são entendidos considerando-se múltiplos espalhamentos, o que equivale a excitar múltiplos átomos de $Hg$.

Um aspecto decisivo do experimento em favor da teoria de Bohr é que fótons com comprimento de onda
\begin{align}
\lambda = \frac{c}{\nu} = \frac{hc}{h\nu} = \frac{hc}{4,9\text{ eV}} \approx 253,03\text{ nm}
\end{align}
são detectados somente para $V_{0}\ge4,9\text{ V}$. Estes são os fótons emitidos pelos átomos de mercúrio quando o elétron retorna do primeiro estado excitado para o estado fundamental.

## Espectro de raios X
Segundo a previsão de Bohr, para um átomo de um elétron temos:
\begin{equation}
E_{n} = -\frac{Z^{2}m_{e}k^{2}e^{4}}{2\hbar^{2}n^{2}}.
\end{equation}
As __frequências de Bohr__ são:
\begin{equation}
\nu_{n_{i},n_{f}} = \frac{|E_{n_{i}}-E_{n_{f}}|}{\hbar}=\frac{Z^{2}m_{e}k^{2}e^{4}}{2\hbar^{2}}\left|\frac{1}{n_{i}^{2}}-\frac{1}{n_{f}^{2}}\right|=cR_{\infty}Z^{2}\left|\frac{1}{n_{i}^{2}}-\frac{1}{n_{f}^{2}}\right|,
\end{equation}
em que definimos a __constante de Rydberg__ como:
\begin{equation}
R_{\infty} := \frac{m_{e}k^{2}e^{4}}{4\pi \hbar^{3}c} \approx 1,1\mathrm{x}10^{7}\text{ m}^{-1}.
\end{equation}
Temos assim que
\begin{equation}
Z = \frac{1}{\sqrt{cR_{\infty}\left|1/n_{i}^{2}-1/n_{f}^{2}\right|}}\sqrt{\nu_{n_{i},n_{f}}}.
\end{equation}
Abaixo mostramos algumas curvas para o "número atômico em função da frequência", que valem para átomos hidrogenóides. Assumindo que transições que produzem raios X são transições para elétrons internos dos átomos, a dependência dos intervalos de energia, e portanto das frequências, com o número atômico será similar à obtida acima para átomos com um único elétron. Os experimentos de __Moseley__ verificaram esses resultados para muitos tipos de átomos diferentes.

```python
(me*k**2*e**4)/(4*math.pi*hb**3*c)
```

```python
import numpy as np
Ri = 10973734.848575922 # m^-1 (constante de Rydberg)
```

```python
def bohrZ(ni,nf,nu):
    return (1/np.sqrt(c*Ri*abs((1/ni**2) - (1/nf**2))))*np.sqrt(nu)
def bohrX():
    import matplotlib
    matplotlib.rcParams.update({'font.size':12}); 
    plt.figure(figsize=(5,10))
    x = np.arange(6,24,0.1); y = bohrZ(1,2,x*10**16); 
    plt.plot(x, y, color = 'blue'); y1 = bohrZ(1,3,x*10**16); 
    plt.plot(x, y1, color = 'cyan'); z = bohrZ(2,3,x*10**16)
    plt.plot(x, z, color = 'green'); z1 = bohrZ(2,4,x*10**16); 
    plt.plot(x, z1, color = 'gray')
    z2 = bohrZ(2,5,x*10**16); plt.plot(x, z2, color = 'red'); 
    plt.xlabel(r'$\sqrt{\nu}(10^{8}Hz^{-1})$')
    plt.legend((r'$K_{\alpha}$',r'$K_{\beta}$',r'$L_{\alpha}$',
                r'$L_{\beta}$',r'$L_{\gamma}$'), loc='upper left')
    plt.ylabel('Z'); plt.show()
bohrX()
```
