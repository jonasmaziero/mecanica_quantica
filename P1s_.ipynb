$1.$ O valor médio de um observável com decomposição espectral $A=\sum_{j=1}^{d}a_{j}|a_{j}\rangle\langle a_{j}|$ de um sistema preparado no estado $|\psi\rangle$ é dado por $\langle A\rangle_{\psi} = \sum_{j=1}^{d}a_{j}Pr(a_{j}|\psi)$. Use a regra de Born para escrever essa expressão como o "sanduíche": $\langle A\rangle_{\psi} = \langle \psi|A|\psi\rangle$.

_Sol._ Da regra de Born teremos
\begin{align}
\langle A\rangle_{\psi} & = \sum_{j=1}^{d}a_{j}Pr(a_{j}|\psi) \\
                        & = \sum_{j=1}^{d}a_{j}|\langle a_{j}|\psi\rangle|^{2} \\
                        & = \sum_{j=1}^{d}a_{j} \langle a_{j}|\psi\rangle^{*}\langle a_{j}|\psi\rangle  \\
                        & = \sum_{j=1}^{d}a_{j} \langle \psi| a_{j}\rangle\langle a_{j}|\psi\rangle  \\
                        & = \langle \psi|\left(\sum_{j=1}^{d}a_{j} | a_{j}\rangle\langle a_{j}|\right)|\psi\rangle  \\
                        & = \langle \psi|A|\psi\rangle.
\end{align}

$2.$ Verifique que $U_{t} = e^{-iHt/\hbar}$ é solução da equação de Schr\"odinger $i\hbar\partial_{t}U_{t}=HU_{t}$.

_Sol._ Faremos isso primeiramente usando $H=\sum_{j}E_{j}|E_{j}\rangle\langle E_{j}|$ e aplicando a função de operador:
\begin{align}
i\hbar\partial_{t} U_{t} & = i\hbar\partial_{t}e^{-iHt/\hbar} \\
                         & = i\hbar\partial_{t}\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = i\hbar\sum_{j}\partial_{t}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = i\hbar\sum_{j}(-i/\hbar)E_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = \sum_{j}e^{-iE_{j}t/\hbar}E_{j}|E_{j}\rangle\langle E_{j}| \\
                         & = \sum_{j}e^{-iE_{j}t/\hbar}H|E_{j}\rangle\langle E_{j}| \\
                         & = H\sum_{j}e^{-iE_{j}t/\hbar}|E_{j}\rangle\langle E_{j}| \\
                         & = H U_{t}.
\end{align}

$3.$ Considere uma partícula de spin 1/2 sujeita a um campo magnético constante na direção $z$ e praparada inicialmente no estado $|z_{+}\rangle$. Qual é o Hamiltoniano desse sistema (forneça sua decomposição espectral)? Qual é o operador de evolução temporal desse sistema (forneça sua decomposição espectral)? Qual é o estado evoluído desse sistema?

_Sol._ Hamiltoniano: 
\begin{equation}
H=-\frac{\hbar\omega}{2}\sigma_{z}.
\end{equation}
Operador de evolução temporal: 
\begin{equation}
U_{t}=e^{-iHt/\hbar}=e^{i\omega\sigma_{z} t/2}=e^{i\omega t/2}|z_{+}\rangle\langle_{+}| + e^{-i\omega t/2}|z_{-}\rangle\langle_{-}|.
\end{equation}
Estado evoluído
\begin{equation}
|\psi_{t}\rangle=U_{t}|\psi_{0}\rangle=U_{t}|z_{+}\rangle=e^{i\omega t/2}|z_{+}\rangle.
\end{equation}

$4.$ Dado que os operadores de quadratura são escritos em termos dos operadores de levantamento e abaixamento como segue:
\begin{align}
X & = \sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}+a) \\
P & = i\sqrt{\frac{\hbar m\omega}{2}}(a^{\dagger}-a),
\end{align}
e que estes últimos possuem a seguinte relação de comutação
\begin{equation}
[a,a^{\dagger}]=\mathbb{I},
\end{equation}
obtenha a relação de comutação entre $X$ e $P$. Nossa incerteza sobre as quadraturas é nula em geral ou não?

_Sol._ Teremos
\begin{align}
[X,P]&=\frac{i\hbar}{2}[a^{\dagger}+a,a^{\dagger}-a] \\ 
&= \frac{i\hbar}{2}([a^{\dagger},a^{\dagger}]-[a^{\dagger},a]+[a,a^{\dagger}]-[a,a])\\
&= i\hbar[a,a^{\dagger}]=i\hbar\mathbb{I}.
\end{align}
Existe um teorema que prova que dois operadores Hermitianos comutam se e somente se compartilham uma mesma base de autovetores. Esse não é o caso das quadraturas. Então podemos preparar o sistema em uma superposição dos autovetores de $X$ medido $P$ e vice versa. Portanto nossa incerteza sobre esses observáveis não é, em geral, não nula.
