# Teoria de perturbação
O objetivo de teoria de perturbação é construir soluções aproximadas (estáticas e/ou dinâmicas) a partir das soluções analíticas obtidas em casos particulares, que geralmente estão relacionados com intensidades de interação maiores em relação às correções que queremos incluir.

## Teoria de perturbação independente do tempo
### Espectros discretos não degenerados
Para Hamiltonianos com espectro discreto não degenerado, começamos considerando que nosso objetivo é obter a solução da equação de autovalores e autovetores:
\begin{equation}
H|\psi_{n}\rangle = E_{n}|\psi_{n}\rangle,
\end{equation}
quando o Hamiltoniano pode ser escrito como 
\begin{equation}
H = H_{0} + H_{1},
\end{equation}
onde $H_{0}$ representa a parte mais intensa da interação e cujos autovalores e autovetores são conhecidos, i.e., conhecemos a solução de 
\begin{equation}
H_{0}|\psi_{n}^{(0)}\rangle = E_{n}^{(0)}|\psi_{n}^{(0)}\rangle,
\end{equation}
e $H_{1}$ é a perturbação, uma interação de intensidade pequena em relação a $H_{0}$. Na sequência obteremos equações que nos possibilitarão, a princípio, calcular correções parturbativas de várias ordens para a energia. Começaremos trocando $H = H_{0} + H_{1}$ por
\begin{equation}
H = H_{0} + \xi H_{1},
\end{equation}
e depois fazemos $\xi=1$. Seguindo, escrevemos a solução geral em termos da não perturbada como segue:
\begin{equation}
|\psi_{n}\rangle = N_{n}(|\psi_{n}^{(0)}\rangle + |\chi_{n}\rangle),
\end{equation}
com $|\chi_{n}\rangle$ não sendo necessariamente normalizado e satisfazendo $\langle\chi_{n}|\psi_{n}^{(0)}\rangle=0$. <br>
__Exercício:__ Verifique que a constate de normalização, considerada $\in\mathbb{R}$, é dada por
\begin{equation}
N_{n} = \frac{1}{\sqrt{1+\langle\chi_{n}|\chi_{n}\rangle}}.
\end{equation}

Subistituindo o Hamiltoniano e autovetors como escritos acima, teremos
\begin{align}
& H|\psi_{n}\rangle = (H_{0}+\xi H_{1})N_{n}(|\psi_{n}^{(0)}\rangle + |\chi_{n}\rangle ) \\
& = N_{n}\left(H_{0}|\psi_{n}^{(0)}\rangle + H_{0}|\chi_{n}\rangle + \xi H_{1}|\psi_{n}^{(0)}\rangle + \xi H_{1}|\chi_{n}\rangle\right) \\
& = N_{n}\left(E_{n}^{(0)}|\psi_{n}^{(0)}\rangle + H_{0}|\chi_{n}\rangle + \xi H_{1}|\psi_{n}^{(0)}\rangle + \xi H_{1}|\chi_{n}\rangle\right) \\
& = E_{n}|\psi_{n}\rangle \\
& = E_{n}N_{n}(|\psi_{n}^{(0)}\rangle + |\chi_{n}\rangle).
\end{align}
Desta equação, prontamente identificamos a relação:
\begin{equation}
(E_{n}-E_{n}^{(0)})|\psi_{n}^{(0)}\rangle = \xi H_{1}|\psi_{n}^{(0)}\rangle + (H_{0}+ \xi H_{1} - E_{n})|\chi_{n}\rangle. \text{ (*)}
\end{equation}
Se tomarmos o produto interno de $|\psi_{n}^{(0)}\rangle$ com ($*$):
\begin{align}
& (E_{n}-E_{n}^{(0)})\langle\psi_{n}^{(0)}|\psi_{n}^{(0)}\rangle = E_{n}-E_{n}^{(0)} \\
& = \xi \langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \langle\psi_{n}^{(0)}|H_{0}|\chi_{n}\rangle + \xi \langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle - E_{n}\langle\psi_{n}^{(0)}|\chi_{n}\rangle \\
& = \xi \langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + E_{n}^{(0)}\langle\psi_{n}^{(0)}|\chi_{n}\rangle + \xi \langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle - 0 \\
& = \xi\left(\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle\right).
\end{align}
ou
\begin{equation}
E_{n} = E_{n}^{(0)} + \xi\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \xi\langle\psi_{n}^{(0)}|H_{1}|\chi_{n}\rangle.  \text{ (**)}
\end{equation}

Note que na última expressão somente último termo está escrito em termos de $|\chi_{n}\rangle$, que não conhecemos. Vamos então decompor este vetor na base $\{|\psi_{n}^{(0)}\rangle\}$:
\begin{align}
|\chi_{n}\rangle & = \sum_{m}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|\chi_{n}\rangle \\
                 & = \sum_{m\ne n}\langle\psi_{m}^{(0)}|\chi_{n}\rangle|\psi_{m}^{(0)}\rangle.
\end{align}
Para obter os coeficientes $\langle\psi_{m}^{(0)}|\chi_{n}\rangle$ para $m\ne n$ tomamos o produto interno de $|\psi_{m}^{(0)}\rangle$ com ($*$):
\begin{align}
& (E_{n}-E_{n}^{(0)})\langle\psi_{m}^{(0)}|\psi_{n}^{(0)}\rangle = 0 \\
& = \xi \langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \langle\psi_{m}^{(0)}|H_{0}|\chi_{n}\rangle + \xi \langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle - E_{n}\langle\psi_{m}^{(0)}|\chi_{n}\rangle \\
& = \xi \langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + E_{m}^{(0)}\langle\psi_{m}^{(0)}|\chi_{n}\rangle + \xi \langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle - E_{n}\langle\psi_{m}^{(0)}|\chi_{n}\rangle.
\end{align}
Assim, como $E_{n}\ne E_{m}^{(0)}$,
\begin{equation}
\langle\psi_{m}^{(0)}|\chi_{n}\rangle = \xi\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}} + \xi\frac{\langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle}{E_{n}-E_{m}^{(0)}}.
\end{equation}
Usando esses coeficientes na decomposição de $|\chi_{n}\rangle$ e o substituindo iterativamente vem que
\begin{align}
|\chi_{n}\rangle & = \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle + \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\chi_{n}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle \\
& = \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle \\ 
& + \xi^{2}\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}}\frac{\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m'}^{(0)}}|\psi_{m}^{(0)}\rangle \\
& + \cdots.
\end{align}

Substituindo essa equação em $(**)$ obteremos
\begin{align}
E_{n} & = E_{n}^{(0)} + \xi\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle \\
& + \xi^{2}\sum_{m\ne n}\frac{\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}-E_{m}^{(0)}} \\
& + \xi^{3}\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{(E_{n}-E_{m}^{(0)})(E_{n}-E_{m'}^{(0)})} \\
& + \cdots.
\end{align}

Agora vamos utilizar a dita _expansão de Rayleigh-Schrödinger_. Notemos que ainda nos resta o $E_{n}$ no denominador das expansões acima. Identificando
\begin{equation}
E_{n} = \sum_{j=0}^{\infty} E_{n}^{(j)}\xi^{j}
\end{equation}
e usando a série $(1-x)^{-1} = \sum_{j=0}^{\infty}x^{j}$ teremos que
\begin{align}
\frac{1}{E_{n}-E_{m}^{(0)}} & = \frac{1}{E_{n}^{(0)}+\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}-E_{m}^{(0)}} \\
& = \frac{1}{(E_{n}^{(0)}-E_{m}^{(0)})}\frac{1}{(1+\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}/(E_{n}^{(0)}-E_{m}^{(0)}))} \\
& = \frac{1}{(E_{n}^{(0)}-E_{m}^{(0)})}\left(1 - \frac{\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}}{E_{n}^{(0)}-E_{m}^{(0)}} + \frac{(\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j})^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} - \cdots\right) \\
& = \frac{1}{E_{n}^{(0)}-E_{m}^{(0)}} - \frac{\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} + \frac{(\sum_{j=1}^{\infty} E_{n}^{(j)}\xi^{j})^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{3}} - \cdots \\
& = \frac{1}{E_{n}^{(0)} - E_{m}^{(0)}} - \xi\frac{E_{n}^{(1)}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} \\
& + \xi^{2}\left( \frac{E_{n}^{(2)}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} + \frac{(E_{n}^{(1)})^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{3}}\right) - \cdots.
\end{align}

Se substituímos esta expressão naquela para $E_{n}$ acima obteremos as várias _ordens de correção para a energia_:
\begin{align}
E_{n} & = \xi^{0}E_{n}^{(0)} + \xi E_{n}^{(1)} + \xi^{2}E_{n}^{(2)} + \xi^{3}E_{n}^{(3)} + \cdots \\
& = \xi^{0}E_{n}^{(0)} + \xi\langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle + \xi^{2}\sum_{m\ne n}\frac{|\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle|^{2}}{E_{n}^{(0)}-E_{m}^{(0)}} \\
& + \xi^{3}\left(\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{(E_{n}^{(0)}-E_{m}^{(0)})(E_{n}^{(0)}-E_{m'}^{(0)})} - E_{n}^{(1)}\sum_{m\ne n}\frac{|\langle\psi_{n}^{(0)}|H_{1}|\psi_{m}^{(0)}\rangle|^{2}}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}} \right) \\
& + \cdots.
\end{align}

Fazendo o análogo para $|\chi_{n}\rangle$ obteremos
\begin{align}
|\chi_{n}\rangle & = \xi\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}^{(0)}-E_{m}^{(0)}}|\psi_{m}^{(0)}\rangle \\ 
& + \xi^{2}\left(\sum_{m\ne n}\sum_{m'\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{m'}^{(0)}\rangle}{E_{n}^{(0)}-E_{m}^{(0)}}\frac{\langle\psi_{m'}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{E_{n}^{(0)}-E_{m'}^{(0)}}|\psi_{m}^{(0)}\rangle - E_{n}^{(1)}\sum_{m\ne n}\frac{\langle\psi_{m}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle}{(E_{n}^{(0)}-E_{m}^{(0)})^{2}}|\psi_{m}^{(0)}\rangle \right) \\
& + \cdots.
\end{align}
_OBS:_ Lembre que antes de utilizar os autoestados $|\psi_{n}\rangle = N_{n}(|\psi_{n}^{(0)}\rangle+|\chi_{n}\rangle)$ para calcular probabilidades, você deve obter a constante de normalização.

### Estrutura fina do átomo de Hidrogênio
Lembrando, com o referencial no centro de massa, o Hamiltoniano do átomo de Hidrogênio tem a forma
\begin{equation}
H_{0} = \frac{P^{2}}{2m}+V = -\frac{\hbar^{2}}{2m}\nabla^{2} - \frac{e^{2}}{4\pi\epsilon_{0}}\frac{1}{r},
\end{equation}
onde $m=m_{e}m_{p}/(m_{e}+m_{p})$. Da solução exata para a equação de Schrödinger temos as energias
\begin{equation}
E_{n}^{(0)} = -\left[\frac{m}{2\hbar^{2}}\left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\right]\frac{1}{n^{2}} = \frac{-13,6\text{ eV}}{n^{2}},
\end{equation}
com $n=1,2,3,\cdots$. Também utilizaremos o raio de Bohr
\begin{equation}
a = \frac{4\pi\epsilon_{0}\hbar^{2}}{me^{2}} \approx 0,528\text{ }\mathring{A}
\end{equation}
e a constante de estrutura fina
\begin{equation}
\alpha = \frac{e^{2}}{4\pi\epsilon_{0}\hbar c} \approx \frac{1}{137,036}.
\end{equation}

#### Correção relativísitica
Lembrando, a energia relativística é dada por
\begin{equation}
T = (\gamma-1)mc^{2},
\end{equation}
com $\gamma = 1/\sqrt{1-v^{2}/c^{2}}$. Ademais, o momento linear relativístico é $P=\gamma mv$. <br>
__Exercício:__ Verifique que
\begin{align}
T & = \sqrt{P^{2}c^{2}+m^{2}c^{4}}-mc^{2} \\ 
  & = mc^{2}\left(\sqrt{1+P^{2}/m^{2}c^{2}} - 1\right).
\end{align}
Seguindo usamos a série $\sqrt{1+x}=1+x/2-x^{2}/8+x^{3}/16-\cdots$ para escrever
\begin{align}
T & = mc^{2}\left(1 + \frac{P^{2}}{2m^{2}c^{2}} - \frac{P^{4}}{8m^{4}c^{4}} + \cdots - 1\right) \\
  & \approx \frac{P^{2}}{2m} - \frac{P^{4}}{8m^{3}c^{2}}.
\end{align}
Então, escrevendo o Hamiltoniano total, nesta aproximação, como $H=H_{0}+H_{1}^{r}$, a correção relativísitca será calculada usando
\begin{equation}
H_{1}^{r}=-\frac{P^{4}}{8m^{3}c^{2}}.
\end{equation}

Correção perturbativa de primeira ordem:
\begin{align}
E_{n}^{(r1)} & = \langle\psi_{n,l,m_{l}}^{(0)}|H_{1}^{r}|\psi_{n,l,m_{l}}^{(0)}\rangle \\
            & = -\frac{1}{8m^{3}c^{2}}\langle\psi_{n,l,m_{l}}^{(0)}|P^{4}|\psi_{n,l,m_{l}}^{(0)}\rangle.
\end{align}
O termo $P^{4}$ pode ser reescrito como segue
\begin{align}
P^{4} & = P^{2}P^{2} \\
      & = 2m(H_{0}-V)2m(H_{0}-V) \\
      & = 4m^{2}\left(H_{0}^{2} - VH_{0} - H_{0}V + V^{2}\right).
\end{align}
Assim
\begin{align}
E_{n}^{(r1)} & = -\frac{1}{2mc^{2}}\left(\langle \psi_{n,l,m_{l}}^{(0)}|H_{0}^{2}|\psi_{n,l,m_{l}}^{(0)}\rangle - \langle\psi_{n,l,m_{l}}^{(0)}|VH_{0}|\psi_{n,l,m_{l}}^{(0)}\rangle - \langle\psi_{n,l,m_{l}}^{(0)}|H_{0}V|\psi_{n,l,m_{l}}^{(0)}\rangle + \langle\psi_{n,l,m_{l}}^{(0)}|V^{2}|\psi_{n,l,m_{l}}^{(0)}\rangle \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} -2E_{n}^{(0)}\langle\psi_{n,l,m_{l}}^{(0)}|V|\psi_{n,l,m_{l}}^{(0)}\rangle + \langle\psi_{n,l,m_{l}}^{(0)}|V^{2}|\psi_{n,l,m_{l}}^{(0)}\rangle \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} +2E_{n}^{(0)}\frac{e^{2}}{4\pi\epsilon_{0}}\langle\psi_{n,l,m_{l}}^{(0)}|r^{-1}|\psi_{n,l,m_{l}}^{(0)}\rangle + \left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\langle\psi_{n,l,m_{l}}^{(0)}|r^{-2}|\psi_{n,l,m_{l}}^{(0)}\rangle \right).
\end{align}
Usando (__exercício__) $\langle\psi_{n,l,m_{l}}^{(0)}|r^{-1}|\psi_{n,l,m_{l}}^{(0)}\rangle=1/an^{2}$ e $\langle\psi_{n,l,m_{l}}^{(0)}|r^{-2}|\psi_{n,l,m_{l}}^{(0)}\rangle=1/a^{2}n^{3}(l+1/2)$ teremos
\begin{align}
E_{n}^{(r1)} & = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} + 2E_{n}^{(0)}\frac{e^{2}}{4\pi\epsilon_{0}}\frac{1}{an^{2}} + \left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\frac{1}{a^{2}n^{3}(l+1/2)} \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} + 2E_{n}^{(0)}\frac{e^{2}}{4\pi\epsilon_{0}}\frac{me^{2}}{4\pi\epsilon_{0}\hbar^{2}}\frac{1}{n^{2}} + \left(\frac{e^{2}}{4\pi\epsilon_{0}}\right)^{2}\frac{m^{2}e^{4}}{(4\pi\epsilon_{0}\hbar^{2})^{2}}\frac{1}{n^{3}(l+1/2)} \right) \\
& = -\frac{1}{2mc^{2}}\left((E_{n}^{(0)})^{2} - 4(E_{n}^{(0)})^{2} + (E_{n}^{(0)})^{2}\frac{4n}{l+1/2} \right) \\
& = -\frac{1}{2mc^{2}}(E_{n}^{(0)})^{2}\left(1 - 4 + \frac{4n}{l+1/2} \right) \\
& = -\frac{(E_{n}^{(0)})^{2}}{2mc^{2}}\left(\frac{4n}{l+1/2} -3 \right).
\end{align}

Com a correção relativística, as energias do átomo de Hidrogênio ficam
\begin{equation}
E_{n} \rightarrow E_{n}^{(0)}\left(1 - \frac{E_{n}^{(0)}}{2mc^{2}}\left(\frac{4n}{l+1/2} -3 \right)\right).
\end{equation}
##### _Ordem de magnitude da correção relativística_:
Considera o caso $n=1$ e $l=0$, para o qual
\begin{align}
E_{1}^{(r1)} & = -\frac{(-13,6\text{ eV})^{2}}{2\mathrm{x}5,11\mathrm{x}10^{5}\text{ eV}}\left(\frac{4\mathrm{x}1}{0+1/2} -3 \right) \\ 
             & = 9,05\mathrm{x}10^{-4}\text{ eV}, 
\end{align}
onde usamos
\begin{align}
mc^{2} & = \frac{m_{e}m_{p}c^{2}}{m_{e}+m_{p}} \\
& = \frac{(9,11\mathrm{x}10^{-31}\text{ kg})(1,67\mathrm{x}10^{-27}\text{ kg})(3\mathrm{x}10^{8}\text{ m/s})}{9,11\mathrm{x}10^{-31}\text{ kg}+1,67\mathrm{x}10^{-27}\text{ kg}} \\
& = 8,18\mathrm{x}10^{-14}\text{ J} = 8,18\mathrm{x}10^{-14}\frac{\text{1 eV}}{1,6\mathrm{x}10^{-19}} \\
& = 5,11\mathrm{x}10^{5}\text{ eV}.
\end{align}

#### Interação spin-órbita
Do referencial do elétron, o próton está em movimento e gera um campo magnético que pode ser calculado usando a lei de Biot-Savart:
\begin{align}
\vec{B} & = \frac{\mu_{0}i}{4\pi}\oint_{c}\frac{d\vec{l}\mathrm{x}\hat{r}}{r^{2}} \\
& = \frac{\mu_{0}i}{4\pi r^{2}}\oint_{c}ds\hat{e} \\
& = \frac{\mu_{0}i}{4\pi r^{2}}2\pi r\hat{e} \\
& = \frac{\mu_{0}e/\tau}{2r}\hat{e},
\end{align}
em que aproximaos a "trajetória" por um círculo. O momento angular do próton, nessas condições, será
\begin{equation}
L = rmv = rm2\pi r/\tau.
\end{equation}
Assim
\begin{align}
\vec{B} & = \frac{\mu_{0}e}{2r}\frac{L}{m2\pi r^{2}}\hat{e} \\
& = \frac{\mu_{0}e}{4\pi mr^{3}}\vec{L} \\
& = \frac{e}{4\pi\epsilon_{0} mc^{2}r^{3}}\vec{L},
\end{align}
onde usamos $c=1/\sqrt{\mu_{0}\epsilon_{0}}$. Este campo magnético interage como o momento magnético de spin do elétron $\vec{\mu}_{e} = (-e/m)\vec{S}_{e}$ segundo o Hamiltoniano:
\begin{align}
H_{1}^{so} & = -\mu_{e}\cdot \vec{B} \\
& = \frac{e^{2}}{8\pi\epsilon_{0} m^{2}c^{2}r^{3}}\vec{S}\cdot\vec{L},
\end{align}
em que introdizimos ainda o fator 1/2 para a precessão de Thomas para o referencial não incercial do elétron.

A correção perturbativa de primeira ordem devido à interação spin-órbita será dada por
\begin{align}
E_{n}^{(so1)} & = \langle\psi_{n,l,m_{l}}^{(0)}|H_{1}^{so}|\psi_{n,l,m_{l}}^{(0)}\rangle \\
            & = \frac{e^{2}}{8\pi\epsilon_{0} m^{2}c^{2}}\langle\psi_{n,l,m_{l}}^{(0)}|\frac{\vec{S}\cdot\vec{L}}{r^{3}}|\psi_{n,l,m_{l}}^{(0)}\rangle.
\end{align}
Para o momento angular total do elétron,
\begin{equation}
\vec{J} = \mathbb{I}_{L}\otimes\vec{S} + \vec{L}\otimes\mathbb{I}_{S},
\end{equation}
teremos que (__exercício__)
\begin{align}
\vec{S}\cdot\vec{L} & = (\mathbb{I}_{L}\otimes\vec{S})\cdot(\vec{L}\otimes\mathbb{I}_{S}) \\ 
& = 2^{-1}\left(J^{2} - \mathbb{I}_{L}\otimes S^{2} - L^{2}\otimes\mathbb{I}_{S} \right).
\end{align}

Assim, trocando $|\psi_{n,l,m_{l}}^{(0)}\rangle = |n,l,s,j,m_{j}\rangle$ teremos
\begin{align}
E_{n}^{(so1)} & = \frac{e^{2}}{8\pi\epsilon_{0} m^{2}c^{2}}\langle n,l,s,j,m_{j}|\frac{2^{-1}\left(J^{2} - \mathbb{I}_{L}\otimes S^{2} - L^{2}\otimes\mathbb{I}_{S} \right)}{r^{3}}|n,l,s,j,m_{j}\rangle \\
& = \frac{e^{2}\hbar^{2}\left(j(j+1)-l(l+1)-s(s+1)\right)}{16\pi\epsilon_{0} m^{2}c^{2}}\langle n,l,s,j,m_{j}|r^{-3}|n,l,s,j,m_{j}\rangle.
\end{align}
Usando $\langle n,l,s,j,m_{j}|r^{-3}|n,l,s,j,m_{j}\rangle = 1/l(l+1/2)(l+1)n^{3}a^{3}$ e $s=1/2$ vem que
\begin{align}
E_{n}^{(so1)} & = \frac{e^{2}\hbar^{2}\left(j(j+1)-l(l+1)-3/4\right)}{16\pi\epsilon_{0} m^{2}c^{2}}\frac{1}{l(l+1/2)(l+1)n^{3}a^{3}} \\
& = \frac{e^{2}\hbar^{2}}{16\pi\epsilon_{0}m^{2}c^{2}}\frac{1}{n^{3}}\frac{m^{3}e^{6}}{(4\pi\epsilon_{0})^{3}\hbar^{6}}\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)} \\
& = \frac{E_{n}^{(0)})^{2}n}{mc^{2}}\left(\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)}\right).
\end{align}
Vemos assim que a correção devido à interação spin-órbita tem a mesma ordem de magnitude que a correção relativística. Vamos então adicionar essas duas contribuições para obter a estrutura fina do átomo de Hidrogênio:
\begin{align}
E_{n} & \rightarrow E_{n}^{(0)} + E_{n}^{r(0)} + E_{n}^{(so0)} \\
& = E_{n}^{(0)} -\frac{(E_{n}^{(0)})^{2}}{2mc^{2}}\left(\frac{4n}{l+1/2} -3 \right) + \frac{E_{n}^{(0)})^{2}n}{mc^{2}}\left(\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)}\right) \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left(n\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)} - \left(\frac{2n}{l+1/2} - 3/2\right)\right) \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left[n\left(\frac{j(j+1)-l(l+1)-3/4}{l(l+1/2)(l+1)} - \frac{2}{l+1/2}\right) + 3/2\right] \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left[n\left(\frac{j(j+1)-l(l+1)-3/4-2l(l+1)}{l(l+1/2)(l+1)}\right) + 3/2\right] \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left[n\left(\frac{j(j+1)-3l(l+1)-3/4}{l(l+1/2)(l+1)}\right) + 3/2\right] \\
& \vdots \\
& = E_{n}^{(0)} + \frac{(E_{n}^{(0)})^{2}}{mc^{2}}\left(3-\frac{4n}{j+1/2}\right) \\
& = E_{n}^{(0)}\left[1 + \frac{\alpha^{2}}{n^{2}}\left(\frac{4n}{j+1/2}-3/4\right)\right].
\end{align}
onde usamos $j=|l-s|,\cdots,l+s = l-1/2,l+1/2$ ou $l=j\mp1/2$.
