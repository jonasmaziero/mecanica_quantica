# Oscilador harmônico 1D

Hamiltoniano
\begin{equation}
H = \frac{P^{2}}{2m} + \frac{m\omega^{2}}{2}X^{2},
\end{equation}
com $[X,P]=i\hbar\mathbb{I}$.

## Diagonalização via operadores escada
Definimos os operadors escada
\begin{align}
& a := \sqrt{\frac{m\omega}{2\hbar}}(X + \frac{i}{m\omega}P) \\
& \Rightarrow a^{\dagger} := \sqrt{\frac{m\omega}{2\hbar}}(X - \frac{i}{m\omega}P),
\end{align}
que nos fornecem
\begin{align}
X & = \sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}+a) \\
P & = i\sqrt{\frac{\hbar m\omega}{2}}(a^{\dagger}-a).
\end{align}

Podemos verificar que
\begin{align}
[a,a^{\dagger}] & = \frac{m\omega}{2\hbar}\left[X + \frac{i}{m\omega}P,X - \frac{i}{m\omega}P\right] \\
& = \frac{m\omega}{2\hbar}\left(\mathbb{0} - \frac{i}{m\omega}[X,P] + \frac{i}{m\omega}[P,X] + \mathbb{0}\right) \\
& = \frac{m\omega}{2\hbar}\frac{-2i}{m\omega}[X,P] \\
& = \frac{-i}{\hbar}i\hbar\mathbb{I} \\
& = \mathbb{I}.
\end{align}
