# Exame

$1.$ Calcule os autovalores e autovetores normalizados da matriz de Pauli $\sigma_{x}=\begin{bmatrix} 0 & 1 \\ 1 & 0 \end{bmatrix}$.

Sol. Autovalores: $\det(\sigma_{x}-\lambda\sigma_{0})=0\Rightarrow \det\begin{bmatrix} -\lambda & 1 \\ 1 & -\lambda \end{bmatrix} \Rightarrow \lambda^{2}-1=0\Rightarrow \lambda=\pm1$. <br>
Autovetores: $(\sigma_{x}-1\sigma_{0})|x_{+}\rangle=|\emptyset\rangle \Rightarrow \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix}\begin{bmatrix} a \\ b \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \Rightarrow a=b=1/\sqrt{2}$, para que $|a|^{2}+|b|^{2}=1$. <br>
$(\sigma_{x}-(-1)\sigma_{0})|x_{-}\rangle=|\emptyset\rangle \Rightarrow \begin{bmatrix} 1 & 1 \\ 1 & 1 \end{bmatrix}\begin{bmatrix} a \\ b \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \Rightarrow a=-b=1/\sqrt{2}$, para que $|a|^{2}+|b|^{2}=1$. Então
\begin{equation}
|x_{+}\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ 1 \end{bmatrix} \text{, } |x_{-}\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ -1 \end{bmatrix}.
\end{equation}

$2.$ Para os estados de polarização da luz (direções de oscilação do campo elétrico) teremos o análogo do que vimos para as componentes do momento magnético intrínseco de uma partícula de spin 1/2, i.e., a base ortonormal de autovetores de $\sigma_{z}$ é $\{|\leftrightarrow\rangle,|\updownarrow\rangle\}$ (polarizações horizontal e vertical), a base ortonormal de autovetores de $\sigma_{x}$ é $\{|\nearrow\rangle=(|\leftrightarrow\rangle+|\updownarrow\rangle)/\sqrt{2},|\nwarrow\rangle=(|\leftrightarrow\rangle-|\updownarrow\rangle)/\sqrt{2}\}$ (polarizações diagonal direita e diagonal esquerda) e a base ortonormal de autovetores de $\sigma_{y}$ é $\{|\circlearrowright\rangle=(|\leftrightarrow\rangle+i|\updownarrow\rangle)/\sqrt{2},|\circlearrowleft\rangle=(|\leftrightarrow\rangle-i|\updownarrow\rangle)/\sqrt{2}\}$ (polarizações circular direita e circular esquerda). Considere analisadores de polarização $P_{x},P_{y},P_{z}$, que fazem o processo de medida análogo ao feito pelos aparelhos de Stern-Gerlach $SG_{x},SG_{y},SG_{z}$, respectivamente. Calcule as probabilidades para os possíveis resultados das seguintes sequências de preparações e medidas:
\begin{align}
(a) \hspace{1cm} & |\leftrightarrow\rangle \hspace{0.3cm} \underrightarrow{P_{z}} \hspace{0.3cm} |?\rangle \text{;}\\
(b) \hspace{1cm} & |\leftrightarrow\rangle \hspace{0.3cm} \underrightarrow{P_{z}} \hspace{0.3cm} |?\rangle \hspace{0.3cm} \underrightarrow{P_{x}} \hspace{0.3cm} |?\rangle \text{;} \\
(c) \hspace{1cm} & |\leftrightarrow\rangle \hspace{0.3cm} \underrightarrow{P_{z}} \hspace{0.3cm} |?\rangle \hspace{0.3cm} \underrightarrow{P_{x}} \hspace{0.3cm} |?\rangle \hspace{0.3cm} \underrightarrow{P_{y}} \hspace{0.3cm} |?\rangle.
\end{align}

_Sol._ (a) Usando a regra de Born, obteremos $\leftrightarrow$ com probabilidade $Pr(\leftrightarrow)=|\langle\leftrightarrow|\leftrightarrow\rangle|^{2}=1$ e $\updownarrow$ com probabilidade $Pr(\updownarrow)=|\langle\leftrightarrow|\updownarrow\rangle|^{2}=0$. <br>
(b) Do item anterior, sabemos que o estado depois da medida $P_{z}$ é $|\leftrightarrow\rangle$. Ao medirmos $P_{x}$ obteremos $\nearrow$ com probabilidade $Pr(\nearrow)=|\langle\leftrightarrow|\nearrow\rangle|^{2}=|\langle\leftrightarrow|(|\leftrightarrow\rangle+|\updownarrow\rangle)/\sqrt{2}|^{2}=1/2$. Para a outra possibilidade, $\nwarrow$, teremos $Pr(\nwarrow)=1-Pr(\nearrow)=1/2$. <br>
(c) Do item anterior temos duas possibilidades de estados pós medida $P_{x}$: $|\nearrow\rangle$ com probabilidade $1/2$ e $|\nwarrow\rangle$ também com probabilidade $1/2$. Para calcular as probabilidades na medida $P_{y}$ podemos usar o operador densidade $\rho=(1/2)|\nearrow\rangle\langle\nearrow|+(1/2)|\nwarrow\rangle\langle\nwarrow|=\sigma_{0}/2$. Teremos assim:
\begin{align}
Pr(\circlearrowright) & = Tr(|\circlearrowright\rangle\langle\circlearrowright|\rho) = Tr(|\circlearrowright\rangle\langle\circlearrowright|\sigma_{0}/2) = Tr(|\circlearrowright\rangle\langle\circlearrowright|)/2=\langle\circlearrowright|\circlearrowright\rangle/2 = 1/2; \\
Pr(\circlearrowleft) & = 1-Pr(\circlearrowright) = 1/2.
\end{align}

$3.$ Considere uma partícula de spin 1/2 preparada no estado $|x_{+}\rangle$ e sujeita a um campo magnético $\vec{B}(t)=B_{0}f_{t}\hat{z}$, com $B_{0}$ sendo uma constante real positiva e $f(t)$ uma função real escalar de $t$. Verifique que o Hamiltoniano $H_{t}=-\hbar\omega f_{t}\sigma_{z}/2$, que depende explicitamente do tempo, comuta em tempos diferentes. Obtenha o estado evoluído no tempo desse quanton, e veritique que este pode ser escrito como $|\psi_{t}\rangle=\cos(\omega F_{t}/2)|x_{+}\rangle+i\sin(\omega F_{t}/2)|x_{-}\rangle$ com $F_{t}:=\int_{0}^{t}f_{t'}dt'$.

_Sol._ Como $[H_{t},H_{t'}]=2^{-1}\hbar\omega(f_{t}\sigma_{z}f_{t'}\sigma_{z}-f_{t'}\sigma_{z}f_{t}\sigma_{z})=2^{-1}\hbar\omega(f_{t}f_{t'}\sigma_{z}\sigma_{z}-f_{t'}f_{t}\sigma_{z}\sigma_{z})=\mathbb{O}\hspace{0.3cm}\forall t,t'$ podemos escrever
\begin{equation}
U_{t} = e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'} = e^{i\omega\int_{0}^{t}f_{t'}dt'\sigma_{z}/2} =: e^{i\omega F_{t}\sigma_{z}/2} = e^{i\omega F_{t}/2}|z_{+}\rangle\langle z_{+}| + e^{-i\omega F_{t}/2}|z_{-}\rangle\langle z_{-}|.
\end{equation}
Assim,
\begin{align}
|\psi_{t}\rangle & = U_{t}|\psi_{0}\rangle = e^{i\omega F_{t}/2}|z_{+}\rangle\langle z_{+}|x_{+}\rangle + e^{-i\omega F_{t}/2}|z_{-}\rangle\langle z_{-}|x_{+}\rangle = (e^{i\omega F_{t}/2}|z_{+}\rangle + e^{-i\omega F_{t}/2}|z_{-}\rangle)/\sqrt{2} \\
& = (\cos(\omega F_{t}/2)(|z_{+}\rangle + |z_{-}\rangle) + i \sin(\omega F_{t}/2)(|z_{+}\rangle - |z_{-}\rangle))/\sqrt{2} \\
& = \cos(\omega F_{t}/2)|x_{+}\rangle + i \sin(\omega F_{t}/2)|x_{-}\rangle.
\end{align}

$4.$ Considere a soma de dois momentos angulares $\vec{J}_{a}$ e $\vec{J}_{b}$ com $j_{a}=1$ e $j_{b}=1$. Obtenha os coeficientes de Clebsch-Gordan para o autovetor $|j_{a}=1,j_{b}=1,j=0,m=0\rangle$ do momento angular total, $\vec{J}=\vec{J}_{a}\otimes\mathbb{I}_{b}+ \mathbb{I}_{a}\otimes\vec{J}_{b}$, escrito em termos da base de autovetores das componentes $z$ dos momentos angulares individuais: $\{|j_{a}=1,m_{a}\rangle\otimes|j_{b}=1,m_{b}\rangle\}$.

_Sol._ Temos $j_{a}=1\Rightarrow m_{a}=-1,0,1$, $j_{b}=1\Rightarrow m_{b}=-1,0,1$, $j=0,1,2$ e $m_{j=0}=0; m_{j=1}=-1,0,1; m_{j=2}=-2,-1,0,1,2$. A restrição $m\ne m_{a}+m_{b}\Rightarrow C_{m_{a}m_{b}m}^{j_{a}j_{b}j}=0$ e a normalização dos estados quânticos nos fornecem
\begin{equation}
|j=2,m=2\rangle = |m_{a}=1\rangle\otimes|m_{b}=1\rangle.
\end{equation}
Usando $J_{-}|j=2,m=2\rangle=\hbar\sqrt{2(2+1)-2(2-1)}|j=2,m=1\rangle = \hbar 2|j=2,m=1\rangle$ e $J_{-}=J_{-}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{-}^{b}$ teremos que
\begin{align}
|j=2,m=1\rangle & = \hbar^{-1}2^{-1}\left( J_{-}^{a}|m_{a}=1\rangle\otimes|m_{b}=1\rangle  + |m_{a}=1\rangle\otimes J_{-}^{b}|m_{b}=1\rangle\right) \\
& = \hbar^{-1}2^{-1}\left( \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=0\rangle\otimes|m_{b}=1\rangle  + \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=1\rangle\otimes|m_{b}=0\rangle\right) \\
& = \left(|m_{a}=0\rangle\otimes|m_{b}=1\rangle  + |m_{a}=1\rangle\otimes|m_{b}=0\rangle\right)/\sqrt{2}.
\end{align}

Aplicando mais uma vez o operado de abaixamento, $J_{-}|j=2,m=1\rangle=\hbar\sqrt{2(2+1)-1(1-1)}|j=2,m=0\rangle = \hbar\sqrt{6}|j=2,m=0\rangle$ obteremos
\begin{align}
|j=2,m=0\rangle & = \hbar^{-1}6^{-1/2}2^{-1/2}\left( J_{-}^{a}|m_{a}=0\rangle\otimes|m_{b}=1\rangle  + |m_{a}=0\rangle\otimes J_{-}^{b}|m_{b}=1\rangle + J_{-}^{a}|m_{a}=1\rangle\otimes|m_{b}=0\rangle  + |m_{a}=1\rangle\otimes J_{-}^{b}|m_{b}=0\rangle \right)  \\
& = \hbar^{-1}12^{-1/2}( \hbar\sqrt{1(1+1)-0(0-1)}|m_{a}=-1\rangle\otimes|m_{b}=1\rangle  + \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=0\rangle\otimes|m_{b}=0\rangle \\ 
& + \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=0\rangle\otimes|m_{b}=0\rangle + \hbar\sqrt{1(1+1)-0(0-1)}|m_{a}=1\rangle\otimes|m_{b}=-1\rangle) \\
& = 6^{-1/2}(|m_{a}=-1\rangle\otimes|m_{b}=1\rangle  + 2|m_{a}=0\rangle\otimes|m_{b}=0\rangle + |m_{a}=1\rangle\otimes|m_{b}=-1\rangle).
\end{align}

Agora vamos para o próximo valor de $j$:
\begin{equation}
|j=1,m=1\rangle = C^{111}_{011}|m_{a}=0\rangle\otimes|m_{b}=1\rangle + C^{111}_{101}|m_{a}=1\rangle\otimes|m_{b}=0\rangle,
\end{equation}
que deve ser ortogonal a $|j=2,m=1\rangle$ e normalizado, o que nos fornece
\begin{equation}
|j=1,m=1\rangle = \left(|m_{a}=0\rangle\otimes|m_{b}=1\rangle  - |m_{a}=1\rangle\otimes|m_{b}=0\rangle \right)/\sqrt{2}.
\end{equation}
Seguindo, usamos $J_{-}|j=1,m=1\rangle =\hbar\sqrt{1(1+1)-1(1-1)}|j=1,m=0\rangle = \hbar\sqrt{2}|j=1,m=0\rangle$ para escrever
\begin{align}
|j=1,m=0\rangle & = \hbar^{-1}2^{-1}(J_{-}^{a}|m_{a}=0\rangle\otimes|m_{b}=1\rangle + |m_{a}=0\rangle\otimes J_{-}^{b}|m_{b}=1\rangle - J_{-}^{a}|m_{a}=1\rangle\otimes|m_{b}=0\rangle - |m_{a}=1\rangle\otimes J_{-}^{b}|m_{b}=0\rangle) \\
& = \hbar^{-1}2^{-1}(\hbar\sqrt{1(1+1)-0(0-1)}|m_{a}=-1\rangle\otimes|m_{b}=1\rangle + \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=0\rangle\otimes|m_{b}=0\rangle \\ 
& - \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=0\rangle\otimes|m_{b}=0\rangle - \hbar\sqrt{1(1+1)-0(0-1)}|m_{a}=1\rangle\otimes|m_{b}=-1\rangle) \\
& = \left(|m_{a}=-1\rangle\otimes|m_{b}=1\rangle  - |m_{a}=1\rangle\otimes|m_{b}=-1\rangle \right)/\sqrt{2}.
\end{align}

Por fim, teremos
\begin{equation}
|j=0,m=0\rangle = C^{110}_{-110}|m_{a}=-1\rangle\otimes|m_{b}=1\rangle + C^{110}_{1-10}|m_{a}=1\rangle\otimes|m_{b}=-1\rangle + C^{110}_{000}|m_{a}=0\rangle\otimes|m_{b}=0\rangle.
\end{equation}
Da ortogonalidade com $|j=1,m=0\rangle$ vemos que $C^{110}_{-110}=C^{110}_{1-10}$. Da ortogonalidade com $|j=2,m=0\rangle$ temos que $C^{110}_{-110}+2C^{110}_{000}+C^{110}_{1-10}=2(C^{110}_{-110}+C^{110}_{000})=0$, que nos fornece $C^{110}_{-110}=-C^{110}_{000}=C^{110}_{1-10}=1/\sqrt{3}$. Assim
\begin{equation}
|j=0,m=0\rangle = (|m_{a}=-1\rangle\otimes|m_{b}=1\rangle + |m_{a}=1\rangle\otimes|m_{b}=-1\rangle - |m_{a}=0\rangle\otimes|m_{b}=0\rangle)/\sqrt{3}.
\end{equation}
