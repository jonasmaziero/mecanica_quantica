# Prova I

$1.$ Mostre que se usarmos a base de autovetores de um operador linear normal $A:\mathbb{C}^{n}\rightarrow\mathbb{C}^{n}$ para representá-lo matricialmente, teremos que $\langle a_{j} |A|a_{k}\rangle = a_{j}\delta_{jk}$, e portanto que a representação produto externo de $A$ é $A=\sum_{j=1}^{n} a_{j}  |a_{j}\rangle\langle a_{j}|$.

_SOL._ Temos que $A|a_{j}\rangle=a_{j}|a_{j}\rangle$. Com isso vem que
\begin{align}
A & = \mathbb{I}A\mathbb{I} = \sum_{j}|a_{j}\rangle\langle a_{j}|A\sum_{k}|a_{k}\rangle\langle a_{k}| = \sum_{j,k}\langle a_{j}|A|a_{k}\rangle|a_{j}\rangle\langle a_{k}| = \sum_{j,k}\langle a_{j}|a_{k}|a_{k}\rangle|a_{j}\rangle\langle a_{k}| \\
& = \sum_{j,k}a_{k}\langle a_{j}|a_{k}\rangle|a_{j}\rangle\langle a_{k}| = \sum_{j,k}a_{k}\delta_{j,k}|a_{j}\rangle\langle a_{k}| = \sum_{j}a_{j}|a_{j}\rangle\langle a_{j}|.
\end{align}

$2.$ Verifique que se o operador Hamiltoniano de um sistema é independente do tempo, então $[U_{t},H_{t}] = 0$. Verifique que neste caso o Hamiltoniano na representação de Heisenberg é equivalente ao Hamiltoniano na representação de Schrödinger.

_SOL._ $H_{t}=H\therefore i\hbar\partial_{t}U_{t}=H_{t}U_{t}=HU_{t}\therefore U_{t}=exp(-iHt/\hbar)$ é solução da equação de Schrödinger. Nesse caso
\begin{equation}
[U_{t},H_{t}]=[e^{-iHt/\hbar},H] = [\sum_{j=0}^{\infty}(-it/\hbar)^{j}H^{j},H]=\sum_{j=0}^{\infty}(-it/\hbar)^{j}[H^{j},H]=\mathbb{O},
\end{equation}
em que usamos que potências de um operador comutam com este operador. <br>
Na representação de Heisenberg
\begin{equation}
H_{t}^{H}=U_{t}^{\dagger}H_{t}^{S}U_{t}=U_{t}^{\dagger}H_{t}U_{t}=U_{t}^{\dagger}U_{t}H_{t} = \mathbb{I}H_{t}=H_{t}^{S}.
\end{equation}

$3.$ Considere uma partícula de spin 1/2 preparada no estado $|n_{+}\rangle = \cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle$, com $\theta\in[0,\pi]$ e $\phi\in[0,2\pi]$. Calcule o estado evoluído sob a ação do operador unitário $U=\frac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix}$. Para este estado evoluído, calcule a probabilidade de obtermos $|z_{+}\rangle$ ao medirmos a componente $z$ do momento magnético dessa partícula.

_SOL._ Temos que
\begin{align}
& U|z_{+}\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix}\begin{bmatrix}1\\0\end{bmatrix} = \frac{1}{\sqrt{2}}\begin{bmatrix}1\\1\end{bmatrix} = (|z_{+}\rangle+|z_{-}\rangle)/\sqrt{2}, \\
& U|z_{-}\rangle = \frac{1}{\sqrt{2}}\begin{bmatrix}1&1\\1&-1\end{bmatrix}\begin{bmatrix}0\\1\end{bmatrix} = \frac{1}{\sqrt{2}}\begin{bmatrix}1\\-1\end{bmatrix} = (|z_{+}\rangle-|z_{-}\rangle)/\sqrt{2}.
\end{align}
Assim
\begin{align}
U|n_{+}\rangle & = U(\cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle) = \cos(\theta/2)U|z_{+}\rangle + e^{i\phi}\sin(\theta/2)U|z_{-}\rangle \\
& = \cos(\theta/2)(|z_{+}\rangle+|z_{-}\rangle)/\sqrt{2} + e^{i\phi}\sin(\theta/2)(|z_{+}\rangle-|z_{-}\rangle)/\sqrt{2} \\ 
&= 2^{-1/2}(\cos(\theta/2)+e^{i\phi}\sin(\theta/2))|z_{+}\rangle + 2^{-1/2}(\cos(\theta/2)-e^{i\phi}\sin(\theta/2))|z_{-}\rangle.
\end{align}
Probabilidade de obtermos $z_{+}$:
\begin{align}
Pr(z_{+}) &= |\langle z_{+}|U|n_{+}\rangle|^{2} = |2^{-1/2}(\cos(\theta/2)+e^{i\phi}\sin(\theta/2))|^{2} \\
& = 2^{-1}(\cos^{2}(\theta/2)+e^{-i\phi}\sin(\theta/2))\cos(\theta/2)+e^{i\phi}\sin(\theta/2))\cos(\theta/2)+\sin^{2}(\theta/2)) \\
& = 2^{-1}(1+2\sin(\theta/2))\cos(\theta/2)\cos\phi).
\end{align}

$4.$ Considere duas partículas de spin 1/2 preparadas no estado $|\Psi\rangle_{AB}=\sqrt{\frac{1}{3}}|z_{+}\rangle_{A}\otimes|z_{-}\rangle_{B} + \sqrt{\frac{2}{3}}|z_{-}\rangle_{A}\otimes|z_{+}\rangle_{B}$. O operador densidade que descreve o spin $A$ é $\rho_{A}=\begin{bmatrix} 1/3&0\\0&2/3\end{bmatrix}$. Calcule o valor médio da componente $z$ do spin A: $\sigma_{z}^{A}=\begin{bmatrix} 1&0\\0&-1 \end{bmatrix}$. Todos os operadores desta questão estão representados na base ortonormal $\{|z_{+}\rangle_{A},|z_{-}\rangle_{A}\}$.

_SOL._ O valor médio é
\begin{equation}
\langle\sigma_{z}^{A}\rangle_{\rho_{A}} = Tr(\rho_A\sigma_{z}^{A}) = Tr(\begin{bmatrix} 1/3&0\\0&2/3\end{bmatrix}\begin{bmatrix} 1&0\\0&-1 \end{bmatrix})=Tr\begin{bmatrix} 1/3&0\\0&-2/3 \end{bmatrix}=1/3-2/3=-1/3.
\end{equation}
