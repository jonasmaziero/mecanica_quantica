$1.$ Dado que a função de onda de uma partícula, confinada em uma dimensão espacial e na base de posição, é $\psi(-w/2<x<w/2)=1/\sqrt{w}\text{, }\psi(x>w/2)=\psi(x<-w/2)=0$, com $w\in\mathbb{R}$ e constante, obtenha a função de onda $\psi(p_{x})$ dessa partícula na base de momento linear.

_SOL._ Está resolvida nas notas de aula, e com o código abaixo.

```python
%run /Users/jonasmaziero/Dropbox/GitHub/mecanica_quantica/init.ipynb
```

```python
def psiP_1d(psi_x,a,b):
    '''retorna a função de onda na base do momento linear, dada psi(x)'''
    x = symbols('x')
    hb = symbols('hbar', constant = True)
    psi_p = integrate(psi_x*cos(px*x/hb),(x,a,b)) -1j*integrate(psi_x*sin(px*x/hb),(x,a,b))
    return (1/sqrt(2*pi*hb))*psi_p
```

```python
w, px = symbols('w p_x', constant = True)
simplify(psiP_1d(1/sqrt(w),-w/2,w/2))
```

$2.$ Verifique que o operador momento angular orbital, $\vec{L}$, comuta tanto com o quadrado do operador posição, $\vec{X}^{2}$, quanto com o quadrado do operador momento linear, $\vec{P}^{2}$.

_SOL._ Usando $\vec{L}=\sum_{j,k,l}\hat{e}_{j}\epsilon_{jkl}X_{k}P_{l}$, $\vec{X}^{2}=\sum_{m}X_{m}^{2}$,  $\vec{P}^{2}=\sum_{m}P_{m}^{2}$, $[X_{j},X_{k}]=\mathbb{O}$, $[P_{j},P_{k}]=\mathbb{O}$ e $[X_{j},P_{k}]=i\hbar\delta_{j,k}\mathbb{I}$, $[AB,C]=A[B,C]+[A,C]B$ e $[A,BC]=B[A,C]+[A,B]C$ obteremos
\begin{align}
[\vec{L},\vec{X}^{2}] & = \left[\sum_{j,k,l}\hat{e}_{j}\epsilon_{jkl}X_{k}P_{l},\sum_{m}X_{m}^{2}\right] = \sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\left[X_{k}P_{l},X_{m}^{2}\right] = \sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\left(X_{k}[P_{l},X_{m}^{2}]+[X_{k},X_{m}^{2}]P_{l}\right) \\
& = \sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\left(X_{k}X_{m}[P_{l},X_{m}]+X_{k}[P_{l},X_{m}]X_{m}+\mathbb{O}P_{l}\right) = 2i\hbar\sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\left(X_{k}X_{m}\delta_{l,m}\right) \\
& = 2i\hbar\sum_{j}\hat{e}_{j}\sum_{k,l}\epsilon_{jkl}X_{k}X_{l} = 2i\hbar\sum_{j}\hat{e}_{j}[X_{k},X_{l}] = \mathbb{O}
\end{align}
e
\begin{align}
[\vec{L},\vec{P}^{2}] & = \left[\sum_{j,k,l}\hat{e}_{j}\epsilon_{jkl}X_{k}P_{l},\sum_{m}P_{m}^{2}\right] = \sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\left[X_{k}P_{l},P_{m}^{2}\right] = \sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\left(X_{k}[P_{l},P_{m}^{2}]+[X_{k},P_{m}^{2}]P_{l}\right) \\
& = \sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\left(X_{k}\mathbb{O}+P_{m}[X_{k},P_{m}]P_{l}+[X_{k},P_{m}]P_{m}P_{l}\right) = 2i\hbar\sum_{j,k,l,m}\hat{e}_{j}\epsilon_{jkl}\delta_{k,m}P_{m}P_{l} \\
& = 2i\hbar\sum_{j}\hat{e}_{j}\sum_{k,l}\epsilon_{jkl}P_{k}P_{l} = 2i\hbar\sum_{j}\hat{e}_{j}[P_{k},P_{l}] = \mathbb{O}.
\end{align}
Nos comutadores acima temos que $\{j,k,l\}$ são permutações de $\{1,2,3\}$.

$3.$ O oscilador de Schwinger é um mapa entre a álgebra do momento angular (AMA) e a álgebra de osciladores harmônicos desacoplados. Esse mapa é definido por:
\begin{align}
& J_{z} := (\hbar/2)(N_{1}\otimes\mathbb{I}_{2} - \mathbb{I}_{1}\otimes N_{2}); \hspace{0.5cm} N_{j} = a_{j}^{\dagger}a_{j} \text{ com } j = 1,2 \\
& J^{2} := (\hbar^{2}/2)N(N/2+\mathbb{I}); \hspace{0.5cm} N = N_{1}\otimes\mathbb{I}_{2}+\mathbb{I}_{1}\otimes N_{2}; \hspace{0.5cm} \mathbb{I}=\mathbb{I}_{1}\otimes\mathbb{I}_{2}, \\
& J_{+} := \hbar a_{1}^{\dagger}\otimes a_{2}; J_{-} := \hbar a_{1}\otimes a_{2}^{\dagger},
\end{align}
com $a_{j}^{\dagger}$ e $a_{j}$  sendo, respectivamente, os operadores de criação e aniquilação do oscilador harmônico $j=1,2$. Verifique que $J_{z}|n_{1}\rangle\otimes|n_{2}\rangle = (\hbar/2)(n_{1}-n_{2})|n_{1}\rangle\otimes|n_{2}\rangle$ e que $J^{2}|n_{1}\rangle\otimes|n_{2}\rangle=(\hbar^{2}/2)(n_{1}+n_{2})((n_{1}+n_{2})/2 + 1)|n_{1}\rangle\otimes|n_{2}\rangle$, em que $n_{j}$ é o número de quanta do oscilador harmônico $j$ e $N_{j}|n_{j}\rangle=n_{j}|n_{j}\rangle$. Use as relações correspondentes da AMA para identificar $j\equiv(n_{1}+n_{2})/2$ e $m\equiv(n_{1}-n_{2})/2$. Usando essa correspondência e os estados obtidos atuando $J_{\pm}$ em $|n_{1}\rangle\otimes|n_{2}\rangle$, verifique que estes operadores não mudam $j$ e que aumentam ou diminuem $m$ de uma unidade.

_SOL._ Começamos por
\begin{align}
J_{z}|n_{1}\rangle\otimes|n_{2}\rangle & = (\hbar/2)(N_{1}\otimes\mathbb{I}_{2}|n_{1}\rangle\otimes|n_{2}\rangle  - \mathbb{I}_{1}\otimes N_{2}|n_{1}\rangle\otimes|n_{2}\rangle ) \\ 
& = (\hbar/2)(N_{1}|n_{1}\rangle\otimes\mathbb{I}_{2}|n_{2}\rangle  - \mathbb{I}_{1}|n_{1}\rangle\otimes N_{2}|n_{2}\rangle ) = (\hbar/2)(n_{1}|n_{1}\rangle\otimes|n_{2}\rangle  - |n_{1}\rangle\otimes n_{2}|n_{2}\rangle ) \\
& = (\hbar/2)(n_{1} - n_{2})|n_{1}\rangle\otimes|n_{2}\rangle.
\end{align}
Seguindo, notamos que
\begin{equation}
N|n_{1}\rangle\otimes|n_{2}\rangle = N_{1}\otimes\mathbb{I}_{2}|n_{1}\rangle\otimes|n_{2}\rangle+\mathbb{I}_{1}\otimes N_{2}|n_{1}\rangle\otimes|n_{2}\rangle = (n_{1}+n_{2})|n_{1}\rangle\otimes|n_{2}\rangle.
\end{equation}
Com isso, vem que
\begin{align}
J^{2}|n_{1}\rangle\otimes|n_{2}\rangle & = (\hbar^{2}/2)N(N/2+\mathbb{I})|n_{1}\rangle\otimes|n_{2}\rangle \\
& = (\hbar^{2}/2)N((n_{1}+n_{2})/2+1)|n_{1}\rangle\otimes|n_{2}\rangle \\
& = (\hbar^{2}/2)(n_{1}+n_{2})((n_{1}+n_{2})/2+1)|n_{1}\rangle\otimes|n_{2}\rangle.
\end{align}
Na AMA temos que $J_{z}|j,m\rangle=\hbar m|j,m\rangle$ e $J^{2}|j,m\rangle = h^{2}j(j+1)|j,m\rangle$. Identificamos assim $|j,m\rangle\equiv |n_{1}\rangle\otimes|n_{2}\rangle$, $m\equiv (n_{1}-n_{2})/2$ e $j = (n_{1}+n_{2})/2$. Agora olhamos para
\begin{align}
J_{+}|n_{1}\rangle\otimes|n_{2}\rangle & = \hbar a_{1}^{\dagger}\otimes a_{2}|n_{1}\rangle\otimes|n_{2}\rangle = \hbar a_{1}^{\dagger}|n_{1}\rangle\otimes a_{2}|n_{2}\rangle \\
& = \hbar\sqrt{n_{1}+1}|n_{1}+1\rangle\otimes\sqrt{n_{2}}|n_{2}-1\rangle = \hbar \sqrt{(n_{1}+1)n_{2}}|n_{1}+1\rangle\otimes|n_{2}-1\rangle, \\
J_{-}|n_{1}\rangle\otimes|n_{2}\rangle & = \hbar a_{1}\otimes a_{2}^{\dagger}|n_{1}\rangle\otimes|n_{2}\rangle \\
& = \hbar\sqrt{n_{1}(n_{2}+1)}|n_{1}-1\rangle\otimes|n_{2}+1\rangle.
\end{align}
Vemos assim que a ação de $J_{+}$ é $n_{1}\rightarrow n_{1}+1$ e $n_{2}\rightarrow n_{2}-1$, que corresponde a $m\rightarrow (n_{1}+1-(n_{2}-1))/2=(n_{1}-n_{2})/2+1=m+1$ e $j\rightarrow(n_{1}+1+(n_{2}-1))/2=(n_{1}+n_{2})/2=j$. Já a ação de $J_{-}$ é $n_{1}\rightarrow n_{1}-1$ e $n_{2}\rightarrow n_{2}+1$, que corresponde a $m\rightarrow (n_{1}-1-(n_{2}+1))/2=(n_{1}-n_{2})/2-1=m-1$ e $j\rightarrow(n_{1}-1+(n_{2}+1))/2=(n_{1}+n_{2})/2=j$. 

$4.$ Considere um sistema de três níveis com Hamiltoniano não perturbado $H_{0} = \begin{bmatrix} a & 0 & 0 \\ 0 & a & 0 \\ 0 & 0 & b \end{bmatrix}$ e sujeito a uma interação perturbativa $H_{1} = \begin{bmatrix} 0 & 0 & c \\ 0 & 0 & d \\ c^{*} & d^{*} & 0 \end{bmatrix}$, com $|c|,|d|\ll b-a>0$. Calcule as correções perturbativas de primeira ordem (caso degenerado). Compare com o resultado exato para os autovalores de $H=H_{0}+H_{1}$, que pode ser visualizado abaixo.

_SOL._ O resultado exato segue abaixo:

```python
H0 = Matrix([[a,0,0],[0,a,0],[0,0,b]]); H = H0 + H1; H.eigenvects()
```

Veremos que as correções perturbativas de primeira ordem, obtidas de $E_{n}^{(1)}c_{n,l} = \sum_{k=1}^{d_{n}}\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle c_{n,k}$, não removem a degenerescência. Abaixo estão mostrados os autovalores de $\tilde{H}_{1} = (\langle\psi_{n,l}^{(0)}|H_{1}|\psi_{n,k}^{(0)}\rangle)=H_{1}$, pois $\{|\psi_{n,k}^{(0)}\rangle\}$ é a base usada para representar $H_{1}$.

```python
a,b = symbols('a,b', real = True)
c,d = symbols('c,d', real = False)
H1 = Matrix([[0,0,c],[0,0,d],[conjugate(c),conjugate(d),0]])
H1t = H1; H1t.eigenvects()
```

Vamos calcular as correções de 2ª ordem, e para isso usamos a eq. $E_{n}^{(2)}c_{n,l} = \sum_{k=1}^{d_{n}}\langle\psi_{n,l}^{(0)}|\tilde{H}_{1}|\psi_{n,k}^{(0)}\rangle c_{n,k}$, com $\tilde{H}_{1}=\sum_{m\ne n}\sum_{j=1}^{d_{m}}\frac{H_{1}|\psi_{m,j}^{(0)}\rangle\langle\psi_{m,j}^{(0)}|H_{1}}{E_{n}^{(0)}-E_{m}^{(0)}}$. Denotamos $|\psi_{1,1}^{(0)}\rangle=|1\rangle$, $|\psi_{1,2}^{(0)}\rangle=|2\rangle$, $|\psi_{2,1}^{(0)}\rangle=|3\rangle$, $E_{1}^{(0)} = a$ com $d_{1}=2$, $E_{2}^{(0)}=b$ com $d_{2}=1$. Assim, usando $H_{1}|1\rangle=c^{*}|3\rangle$, $H_{1}|2\rangle=d^{*}|3\rangle$ e $H_{1}|3\rangle=c|1\rangle+d|2\rangle$ obteremos
\begin{align}
\tilde{H}_{1} & = \sum_{j=1}^{d_{2}}\frac{H_{1}|\psi_{2,j}^{(0)}\rangle\langle\psi_{2,j}^{(0)}|H_{1}}{E_{1}^{(0)}-E_{2}^{(0)}} + \sum_{j=1}^{d_{1}}\frac{H_{1}|\psi_{1,j}^{(0)}\rangle\langle\psi_{1,j}^{(0)}|H_{1}}{E_{2}^{(0)}-E_{1}^{(0)}} \\
& = \frac{H_{1}|\psi_{2,1}^{(0)}\rangle\langle\psi_{2,1}^{(0)}|H_{1}}{a-b} + \frac{H_{1}|\psi_{1,1}^{(0)}\rangle\langle\psi_{1,1}^{(0)}|H_{1}+H_{1}|\psi_{1,2}^{(0)}\rangle\langle\psi_{1,2}^{(0)}|H_{1}}{b-a} \\
& = -\frac{H_{1}|3\rangle\langle 3|H_{1}}{b-a} + \frac{H_{1}|1\rangle\langle 1|H_{1}+H_{1}|2\rangle\langle 2|H_{1}}{b-a} \\
& = -\frac{(c|1\rangle+d|2\rangle)(c^{*}\langle 1|+d^{*}\langle 2|)}{b-a} + \frac{(|c|^{2}+|d|^{2})}{b-a}|3\rangle\langle 3| \\
& = -\frac{|c|^{2}|1\rangle\langle 1|+cd^{*}|1\rangle\langle 2|+dc^{*}|2\rangle\langle 1|+|d|^{2}|2\rangle\langle 2|}{b-a} + \frac{(|c|^{2}+|d|^{2})}{b-a}|3\rangle\langle 3| \\ 
& \doteq \frac{1}{b-a}\begin{bmatrix} -|c|^{2} & -cd^{*} & 0 \\ -dc^{*} & -|d|^{2} & 0 \\ 0 & 0 & |c|^{2}+|d|^{2} \end{bmatrix}.
\end{align}

```python
H1t = (1/(b-a))*Matrix([[-abs(c)**2,-c*conjugate(d),0],[-d*conjugate(c),-abs(d)**2,0],[0,0,abs(c)**2+abs(d)**2]])
H1t
simplify(H1t.eigenvects())
```
