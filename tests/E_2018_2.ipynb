$1.$ Considere um sistema quântico de três níveis em que a ação do operador Hamiltoniano na base ortonormal $\{|1\rangle,|2\rangle,|3\rangle\}$ é
\begin{align}
& H|1\rangle=\Omega|1\rangle+\Omega|3\rangle, \\
& H|2\rangle=\Omega|2\rangle+\Delta|3\rangle, \\
& H|3\rangle=\Omega|1\rangle+\Delta|2\rangle+\Omega|3\rangle.
\end{align}
Forneça a representação matricial de $H$ na base acima. Calcule os autovalores de $H$. Se o sistema foi preparado no estado $|\psi\rangle=(\Delta/\sqrt{\Omega^{2}+\Delta^{2}})|1\rangle-(\Omega/\sqrt{\Omega^{2}+\Delta^{2}})|2\rangle$, qual é a probabilidade de em uma medida de energia obtermos o valor $\Omega$?

_Sol._ Olhando para $\langle j|H|k\rangle$, e usando a ação acima, obtemos
\begin{equation}
H = \begin{bmatrix}
\Omega & 0 & \Omega \\ 0 & \Omega & \Delta \\ \Omega & \Delta & \Omega
\end{bmatrix}.
\end{equation}
Os autovalores são calculados resolvendo a equação característica 
\begin{equation}
\det\begin{bmatrix}
\Omega-E & 0 & \Omega \\ 0 & \Omega-E & \Delta \\ \Omega & \Delta & \Omega-E
\end{bmatrix}=0,
\end{equation}
e são dados por $E_{-}=\Omega-\sqrt{\Omega^{2}+\Delta^{2}}\text{, }E_{0}=\Omega\text{, }E_{+}=\Omega+\sqrt{\Omega^{2}+\Delta^{2}}$. O autovetor normalizado correspondente ao autovalor $\Omega$ pode ser escrito como
$|\Omega\rangle=\begin{bmatrix}-\Delta/\sqrt{\Omega^{2}+\Delta^{2}} & \Omega/\sqrt{\Omega^{2}+\Delta^{2}} & 0 \end{bmatrix}^{T}$. Notando que $|\Omega\rangle=-|\psi\rangle$, vemos que $Pr(E=\Omega)=|\langle\Omega|\psi\rangle|^{2}=1$.

$2.$ Para o mesmo sistema da equação anterior, com Hamiltoniano independente do tempo, dado que os autovetores de $H$ são 
\begin{align}
& |E_{-}\rangle=(1/\sqrt{2})\begin{bmatrix}-\Omega/\sqrt{\Omega^{2}+\Delta^{2}} & -\Delta/\sqrt{\Omega^{2}+\Delta^{2}} & 1 \end{bmatrix}^{T}, \\
& |E_{0}\rangle=\begin{bmatrix}-\Delta/\sqrt{\Omega^{2}+\Delta^{2}} & \Omega/\sqrt{\Omega^{2}+\Delta^{2}} & 0 \end{bmatrix}^{T}, \\
& |E_{+}\rangle=(1/\sqrt{2})\begin{bmatrix}\Omega/\sqrt{\Omega^{2}+\Delta^{2}} & \Delta/\sqrt{\Omega^{2}+\Delta^{2}} & 1 \end{bmatrix}^{T},
\end{align}
e dito que o estado inicial do sistema é o $|\psi\rangle$ da questão anterior, obtenha o estado evoluído no tempo do sistema.

_Sol._ Nesse caso
\begin{align}
|\psi_{t}\rangle & = U_{t}|\psi\rangle = e^{-iHt/\hbar}|\psi\rangle \\
& = (e^{-iE_{-}t/\hbar}|E_{-}\rangle\langle E_{-}|+e^{-iE_{0}t/\hbar}|E_{0}\rangle\langle E_{0}|+e^{-iE_{+}t/\hbar}|E_{+}\rangle\langle E_{+}|)(-|E_{0}\rangle) \\
& = 0-e^{-iE_{0}t/\hbar}|E_{0}\rangle\langle E_{0}|E_{0}\rangle+0 \\
& \propto |\psi\rangle.
\end{align}

$4.$ Considerando o momento angular em MQ, dado que $J^{2}|j,m\rangle=\hbar^{2}j(j+1)|j,m\rangle$, $J_{3}|j,m\rangle=\hbar m|j,m\rangle$, que os operadores escada são definidos como $J_{\pm}=J_{1}\pm iJ_{2}$,  e que $J_{\pm}|j,m\rangle\propto|j,m\pm1\rangle$, e ainda dada a relação de comutação $[J_{j},J_{k}]=i\hbar\epsilon_{jkl}J_{l}$, mostre que
\begin{equation}
J_{\pm}|j,m\rangle=\hbar\sqrt{j(j+1)-m(m\pm1)}|j,m\pm1\rangle.
\end{equation}

_Sol._ Se escrevemos $J_{\pm}|j,m\rangle=c|j,m\pm1\rangle$ então o produto interno
\begin{align}
& (J_{\pm}|j,m\rangle,J_{\pm}|j,m\rangle)=|c|^{2}(|j,m\rangle,|j,m\rangle)=|c|^{2} = \langle j,m|J_{\pm}^{\dagger}J_{\pm}|j,m\rangle = \langle j,m|J_{\mp}J_{\pm}|j,m\rangle \\ 
& = \langle j,m|(J_{1}^{2}+J_{2}^{2}\mp i[j_{1},J_{2}])|j,m\rangle = \langle j,m|(J^{2}-J_{3}^{2}\pm \hbar J_{3})|j,m\rangle =\langle j,m|(\hbar^{2}j(j+1)-\hbar^{2}m^{2}\pm\hbar^{2} m)|j,m\rangle \\ 
& = \hbar^{2}(j(j+1)-m(m\pm 1))\langle j,m|j,m\rangle,
\end{align}
que, assumindo $c\in\mathbb{R}$, nos fornece o resultado buscado.
