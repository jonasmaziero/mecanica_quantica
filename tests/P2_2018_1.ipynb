$1.$ A função de onda que nos fornece o pacote de ondas Gaussiano em uma dimensão lê-se:
\begin{equation}
\psi(x) = \frac{1}{\pi^{1/4}d^{1/2}}e^{ikx-x^{2}/2d^{2}},
\end{equation}
com $k,d\in\mathbb{R}$. Para este estado, calcule o desvio padrão da posição e do momento linear. Comente sobre o que o resultado obtido tem de notável com relação à relação de incerteza posição-momento de Heisenberg. Use $\int_{-\infty}^{+\infty}e^{-ax^{2}}dx=\sqrt{\pi/a}$, $\int_{-\infty}^{+\infty}xe^{-ax^{2}}dx=0$ e $\int_{-\infty}^{+\infty}x^{2}e^{-ax^{2}}dx=\pi^{1/2}/2a^{3/2}$.

_Sol._ Lembrando, para $n\in\mathbb{N_{>0}}$ temos $\langle X^{n}\rangle = \int_{-\infty}^{\infty}x^{n}|\psi(x)|^{2}dx$ e $\langle P^{n}\rangle = \int_{-\infty}^{\infty}\psi^{*}(x)(-i\hbar\partial_{x})^{n}\psi(x)dx$. Assim
\begin{align}
\langle X\rangle & = \int_{-\infty}^{+\infty}x|\psi(x)|^{2}dx \\
& = \frac{1}{\pi^{1/2}d}\int_{-\infty}^{+\infty}xe^{ikx-x^{2}/2d^{2}}e^{-ikx-x^{2}/2d^{2}}dx \\
& = \frac{1}{\pi^{1/2}d}\int_{-\infty}^{+\infty}xe^{-x^{2}/d^{2}}dx \\
& = 0
\end{align}
pois a função sendo integrada é ímpar.

Para o quadrado da posição usa o truque de Feynman:
\begin{align}
\langle X^{2}\rangle & = \int_{-\infty}^{+\infty}x^{2}|\psi(x)|^{2}dx \\
& = \frac{1}{\pi^{1/2}d}\int_{-\infty}^{+\infty}x^{2}e^{-x^{2}/d^{2}}dx \\
& = -\frac{1}{\pi^{1/2}d}\partial_{d^{-2}}\int_{-\infty}^{+\infty}e^{-d^{-2}x^{2}}dx \\
& = -\frac{1}{\pi^{1/2}d}\partial_{d^{-2}}\sqrt{\pi/d^{-2}} \\
& = -\frac{1}{d}\partial_{d^{-2}}(d^{-2})^{-1/2} \\
& = -\frac{1}{d}(-1/2)(d^{-2})^{-3/2} \\
& = \frac{d^{2}}{2}.
\end{align}

Para o momento linear
\begin{align}
\langle P\rangle & = \int_{-\infty}^{+\infty}\psi^{*}(x)(-i\hbar\partial_{x})\psi(x)dx \\
& = \frac{-i\hbar}{\pi^{1/2}d}\int_{-\infty}^{+\infty}e^{-ikx-x^{2}/2d^{2}}\partial_{x}e^{ikx-x^{2}/2d^{2}}dx \\
& = \frac{-i\hbar}{\pi^{1/2}d}\int_{-\infty}^{+\infty} (ik-x/d^{2})e^{-ikx-x^{2}/2d^{2}}e^{ikx-x^{2}/2d^{2}}dx \\
& = \frac{-i\hbar}{\pi^{1/2}d}\left( ik\int_{-\infty}^{+\infty}e^{-d^{-2}x^{2}}dx   -d^{-2}\int_{-\infty}^{+\infty}xe^{-x^{2}/d^{2}}dx \right) \\
& = \frac{-i\hbar}{\pi^{1/2}d}\left( ik\sqrt{\pi/d^{-2}}-0 \right) \\
& = \hbar k.
\end{align}

Para o quadrado do momento linear
\begin{align}
\langle P^{2}\rangle & = \int_{-\infty}^{+\infty}\psi^{*}(x)(-i\hbar\partial_{x})^{2}\psi(x)dx \\
& = \frac{-\hbar^{2}}{\pi^{1/2}d}\int_{-\infty}^{+\infty}e^{-ikx-x^{2}/2d^{2}}\partial_{xx}e^{ikx-x^{2}/2d^{2}}dx \\
& = \frac{-\hbar^{2}}{\pi^{1/2}d}\int_{-\infty}^{+\infty}e^{-ikx-x^{2}/2d^{2}}\partial_{x}\left((ik-d^{-2}x)e^{ikx-x^{2}/2d^{2}}\right)dx \\
& = \frac{-\hbar^{2}}{\pi^{1/2}d}\int_{-\infty}^{+\infty}e^{-ikx-x^{2}/2d^{2}}\left(-d^{-2}e^{ikx-x^{2}/2d^{2}} + (ik-d^{-2}x)^{2}e^{ikx-x^{2}/2d^{2}}\right)dx \\
& = \frac{-\hbar^{2}}{\pi^{1/2}d}\left(-(d^{-2}+k^{2})\int_{-\infty}^{+\infty}e^{-d^{-2}x^{2}}dx + 2ikd^{-2}\int_{-\infty}^{+\infty}xe^{-d^{-2}x^{2}}dx + d^{-4}\int_{-\infty}^{+\infty}x^{2}e^{-d^{-2}x^{2}}dx\right) \\
& = \frac{-\hbar^{2}}{\pi^{1/2}d}\left(-(d^{-2}+k^{2})\sqrt{\pi/d^{-2}} + 0 + d^{-4}\pi^{1/2}2^{-1}(d^{-2})^{-3/2} \right) \\
& = -\hbar^{2}d^{-1}\left(-(d^{-2}+k^{2})d + d^{-1}2^{-1} \right) \\
& = \hbar^{2}k^{2} + \hbar^{2}d^{-2}2^{-1}.
\end{align}

As variâncias são $\mathrm{Var}(X)=\langle X^{2}\rangle-\langle X\rangle^{2} = 2^{-1}d^{2}$ e $\mathrm{Var}(P)=\langle P^{2}\rangle-\langle P\rangle^{2} = \hbar^{2}d^{-2}2^{-1}$. Isso nos leva a $\mathrm{Var}(X)\mathrm{Var}(P)=\hbar^{2}2^{-2}$, que é o mínimo possível na relação de incerteza de Heisenberg.

Abaixo mostro como essas contas podem ser facilmente verificadas com o Sympy.

```python
from sympy import *
init_printing(use_unicode=True)
x, k = symbols('x k', real=True)
d = symbols('d', real=True, positive=True)
a = symbols('a', real=True, positive=True)
```

```python
# integral Gaussiana
integrate(exp(-a*x**2),(x,-oo,+oo))
```

```python
def psi(x,k,d):
    return (1/(pi**(1/4)*d**(1/2)))*exp(I*k*x-x**2/(2*d**2))
def psic(x,k,d):
    return (1/(pi**(1/4)*d**(1/2)))*exp(-I*k*x-x**2/(2*d**2))
psi(x,k,d)
```

```python
def xavg():
    return integrate(x*psic(x,k,d)*psi(x,k,d),(x,-oo,+oo))
simplify(xavg())
```

```python
def x2avg():
    return integrate(x**2*psic(x,k,d)*psi(x,k,d),(x,-oo,+oo))
simplify(x2avg())
```

```python
hb = symbols('hbar', real=True, positive=True)
def pavg():
    return integrate(psic(x,k,d)*(-I*hb)*diff(psi(x,k,d),x),(x,-oo,+oo))
simplify(pavg())
```

```python
def p2avg():
    return integrate(psic(x,k,d)*(-I*hb)**2*diff(psi(x,k,d),x,2),(x,-oo,+oo))
simplify(p2avg())
```

$2.$ Mostre que a função de onda na base de momento linear correspondente à função de onda na base de posição da questão anterior é
\begin{equation}
\psi(p) = \frac{d^{1/2}}{\hbar^{1/2}\pi^{1/4}}e^{-d^{2}(p-\hbar k)^{2}/2\hbar^{2}}.
\end{equation}
Use $\langle x|p\rangle = (2\pi\hbar)^{-1/2}e^{ipx/\hbar}$ e complete quadrados para resolver a integral envolvida nos cálculos.

_Sol._ O essencial aqui é notar que
\begin{align}
\psi(p) & = \langle p|\psi\rangle \\
& = \langle p|\mathbb{I}|\psi\rangle \\
& = \int_{-\infty}^{+\infty}\langle p|x\rangle\langle x|\psi\rangle dx \\
& = \int_{-\infty}^{+\infty}\langle p|x\rangle\psi(x)dx.
\end{align}
Mas se olhamos para
\begin{align}
\langle x|P|p\rangle & = \langle x|p|p\rangle \\
-i\hbar\partial_{x}\langle x|p\rangle & = p\langle x|p\rangle,
\end{align}
vemos que os coeficientes na combinação linear acima podem ser obtidos de $\langle x|p\rangle = Ce^{ipx/\hbar}$. Agora
\begin{align}
\langle x|x'\rangle & = \int_{-\infty}^{+\infty}\langle x|p\rangle\langle p|x'\rangle dp \\
\delta(x-x') & = \int_{-\infty}^{+\infty}Ce^{ipx/\hbar}C^{*}e^{-ipx'/\hbar} dp\\
& = |C|^{2}\hbar 2\pi(2\pi)^{-1}\int_{-\infty}^{+\infty}e^{ip(x-x')/\hbar} dp/\hbar \\
& = |C|^{2}\hbar 2\pi\delta(x-x').
\end{align}
Usamos $C\in\mathbb{R}$ para obter $\langle x|p\rangle = (2\pi\hbar)^{-1/2}e^{ipx/\hbar}$. Com isso teremos
\begin{align}
\psi(p) & = (2\pi\hbar)^{-1/2}\int_{-\infty}^{+\infty}e^{-ipx/\hbar}\psi(x)dx \\
& = (2\pi\hbar)^{-1/2}\pi^{-1/4}d^{-1/2}\int_{-\infty}^{+\infty}e^{-\hbar^{-1}ipx}e^{ikx-2^{-1}d^{-2}x^{2}}dx \\
& = (2\hbar)^{-1/2}\pi^{-3/4}d^{-1/2}\int_{-\infty}^{+\infty}e^{-i\hbar^{-1}(p-\hbar k)x-2^{-1}d^{-2}x^{2}}dx \\
& = (2\hbar)^{-1/2}\pi^{-3/4}d^{-1/2}\int_{-\infty}^{+\infty}e^{-\left(2^{-1/2}di\hbar^{-1}(p-\hbar k)+2^{-1/2}d^{-1}x\right)^2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}dx,
\end{align}
em que completamos quadrados para obter a última expressão. Fazendo a seguinte mudança de variável: $y = 2^{-1/2}di\hbar^{-1}(p-\hbar k)+2^{-1/2}d^{-1}x$ obtemos
\begin{align}
\psi(p) & = (2\hbar)^{-1/2}\pi^{-3/4}d^{-1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}\int_{-\infty}^{+\infty}e^{-y^{2}}dy2^{1/2}d \\
& = \hbar^{-1/2}\pi^{-3/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}\pi^{1/2} \\
& = \hbar^{-1/2}\pi^{-1/4}d^{1/2}e^{-2^{-1}d^{2}\hbar^{-2}(p-\hbar k)^{2}}.
\end{align}

Na sequência verificamos esse resultado usando o Sympy.

```python
p = symbols('p', real=True)
def psip(p):
    return integrate((1/sqrt(2*pi*hb))*exp(-I*p*x/hb)*psi(x,k,d),(x,-oo,+oo))
simplify(psip(p))
```

```python
# verificação da condição de normalização
integrate(psip(p)**2,(p,-oo,+oo))
```

$3.$ Considere a soma de dois momentos angulares $\vec{J}_{a}$ e $\vec{J}_{b}$ com $j_{a}=j_{b}=1$. Liste os possíveis valores de $m_{a}$, $m_{b}$, $j$ e $m$. Obtenha os coeficientes de Clebsch-Gordan para o estado $|j=2,m=1\rangle$. Para um sistema quântico preparado neste estado, qual é a probabilidade de obtermos a componente $z$ de $\vec{J}_{a}$ nula?

_Sol_. Como $m_{s}=-j_{s},-j_{s}+1,\cdots,j_{s}$ para $s=a,b$, temos que $m_{s}=-1,0,1$. Por outro lado, $j=|j_{a}-j_{b}|,|j_{a}-j_{b}|+1,j_{a}+j_{b}$, i.e., $j=0,1,2$. Assim $m(j=0)=0$, $m(j=1)=-1,0,1$ e $m(j=2)=-2,-1,0,1,2$. Agora lembremos que $|j,m\rangle=\sum_{m_{a}=-j_{a}}^{j_{a}}\sum_{m_{b}=-j_{b}}^{j_{b}}C_{m_{a}m_{b}m}^{j_{a}j_{b}j}|m_{a}\rangle\otimes|m_{b}\rangle$ com $m=m_{a}+m_{b}$ necessário para que os coeficientes possam ser não nulos. Vemos assim que $|j=2,m=2\rangle=|m_{a}=1\rangle\otimes|m_{b}=1\rangle$. Ademais $J_{-}|j=2,m=2\rangle = \hbar\sqrt{2(2+1)-2(2-1)}|j=2,m=2-1\rangle = \hbar\sqrt{4}|j=2,m=1\rangle$. Ou seja,
\begin{align}
|j=2,m=1\rangle & = (2\hbar)^{-1} J_{-}|j=2,m=2\rangle \\
& = (2\hbar)^{-1}\left(J_{-}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{-}^{b}\right)|m_{a}=1\rangle\otimes|m_{b}=1\rangle \\
& = (2\hbar)^{-1}\left(\hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=1-1\rangle\otimes|m_{b}=1\rangle + \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=1\rangle\otimes|m_{b}=1-1\rangle \right) \\
& = 2^{-1/2}\left(|m_{a}=0\rangle\otimes|m_{b}=1\rangle + |m_{a}=1\rangle\otimes|m_{b}=0\rangle \right).
\end{align}
A probabilidade pode ser calculada pelo módulo quadrado do coeficiente correspondente, que neste caso é $1/2$.

$4.$ Qual é a expressão geral (inicial) para a correção perturbativa de primeira ordem para estados não degenerados $|\psi_{n}^{(0)}\rangle$? A forma aproximada para o Hamiltoniano de perturbação para a correção relativística é $H_{1}^{r}=-P^{4}/8m^{3}c^{2}$. Dado que para o átomo de Hidrogênio a expressão final para essa correção é \begin{equation} E_{n}^{(r1)} = -\frac{(E_{n}^{(0)})^{2}}{2mc^{2}}\left(\frac{4n}{l+1/2} -3 \right), \end{equation} obtenha seus valores para $n=2$. Use $mc^{2}\approx 5,12\mathrm{x}10^{5}\text{ eV}$.

_Sol._ A correção de primeira ordem é o valor médio do Hamiltoniano de perturbação $H_{1}$ no autovetor do Hamiltoniano não perturbado $H_{0}$ considerado: \begin{equation} E_{n}^{(1)} = \langle\psi_{n}^{(0)}|H_{1}|\psi_{n}^{(0)}\rangle. \end{equation} <br>
Pata $n=2$ podemos ter $l=0$ e $l=1$. Ademais, $E_{n}^{(0)}=-13,6\text{ eV}/n^{2}=-3,4\text{ eV}$ e $mc^{2} \approx 5.12\mathrm{x}10^{5}\text{ eV}$. Como mostrado abaixo, para $l=0$ a correção é $1,47\mathrm{x}10^{-4}\text{ eV}$ e para $l=1$ é $2,63\mathrm{x}10^{-5}\text{ eV}$.

```python
me, mp, cc, ee = symbols('m_{e} m_{p} c e')
me = 9.11*10**-31
mp = 1.67*10**-27
cc = 3*10**8
ee = 1.6*10**-19
def m(ma, mb):
    return ma*mb/(ma+mb)
def mc2():
    return m(me,mp)*cc**2/(ee)
mc2()
```

```python
def En(n):
    return -13.6/n**2
En(2)
```

```python
def Enr1(n,l):
    return -((En(2)**2)/(2*mc2()))*((4*n)/(l+1/2)-3)
Enr1(2,0)
```

```python
Enr1(2,1)
```
