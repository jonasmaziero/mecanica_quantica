$1.$ Escreva explicitamente a relação de incerteza de Robertson para as componentes do momento angular, e a discuta no que concerne a (in)compatibilidade dessas componentes.

_SOL._ Usando a relação de comutação que define a álgebra do momento angular, $[L_{j},L_{k}]=i\hbar\epsilon_{jkl}L_{l},$ e a RIR, $\sigma_{A}\sigma_{B}\ge 2^{-1}|\langle[A,B]\rangle|,$ teremos que
\begin{align}
\sigma_{L_{j}}\sigma_{L_{k}} & \ge 2^{-1}|\langle[L_{j},L_{k}]\rangle| = 2^{-1}|\langle i\hbar\epsilon_{jkl}L_{l}\rangle| = 2^{-1}\hbar\epsilon_{jkl}|\langle L_{l}\rangle|.
\end{align}
Assumindo $j\ne k\ne l$, vemos assim que se $\langle L_{l}\rangle\ne0$ então o produto das indeterminações de $L_{j}$ e $L_{k}$ não pode ser nulo. Ou seja, em geral, as componentes do momento angular são observáveis incompatíveis.

$2.$ Verifique que $[L^{2},L_{2}]=[L^{2},L_{3}]=0$.

_SOL._ Como foi mostrado nas notas de aula, $[L^{2},L_{k}] = i\hbar\sum_{j,l=1}^{3}\varepsilon_{jkl}\{L_{j},L_{l}\}.$ Assim
\begin{align}
[L^{2},L_{2}] & = i\hbar\sum_{j,l=1}^{3}\varepsilon_{j2l}\{L_{j},L_{l}\} = i\hbar(\varepsilon_{123}\{L_{1},L_{3}\} + \varepsilon_{321}\{L_{3},L_{1}\}) \\
& = i\hbar(\varepsilon_{123}\{L_{1},L_{3}\} + \varepsilon_{321}\{L_{1},L_{3}\}) = i\hbar\{L_{1},L_{3}\}(\varepsilon_{123} + \varepsilon_{321}) \\
& = i\hbar\{L_{1},L_{3}\}(\varepsilon_{123} - \varepsilon_{123}) \\ 
& = 0, \\
[L^{2},L_{3}] & = i\hbar\sum_{j,l=1}^{3}\varepsilon_{j3l}\{L_{j},L_{l}\} = i\hbar(\varepsilon_{132}\{L_{1},L_{2}\} + \varepsilon_{231}\{L_{2},L_{1}\}) \\
&  = i\hbar(\varepsilon_{132}\{L_{1},L_{2}\} + \varepsilon_{231}\{L_{1},L_{2}\}) = i\hbar\{L_{1},L_{2}\}(\varepsilon_{132} + \varepsilon_{231}) \\
&  = i\hbar\{L_{1},L_{2}\}(\varepsilon_{132} + \varepsilon_{231}) = i\hbar\{L_{1},L_{2}\}(\varepsilon_{132} - \varepsilon_{132}) \\
& = 0.
\end{align}

$3.$ Lembrando, $L|l,m_{l}\rangle:=\lambda|l,m_{l}\rangle$ e $L_{3}|l,m_{l}\rangle:=\hbar m_{l}|l,m_{l}\rangle$. Analogamente ao que fizemos em aula usando $L^{2}=L_{-}L_{+} + L_{3}^{2} + \hbar L_{3}$, obtenha $\lambda=\hbar^{2}l(l+1)$ usando $L^{2}=L_{+}L_{-} + L_{3}^{2} - \hbar L_{3}$.

_SOL._ Escolhendo o autovetor $|l,-l\rangle$ de $L^{2}$ teremos
\begin{align}
& L^{2}|l,-l\rangle = \lambda|l,-l\rangle \\
&= (L_{+}L_{-} + L_{3}^{2} - \hbar L_{3})|l,-l\rangle = |\oslash\rangle + (\hbar^{2}(-l)^{2}-\hbar^{2}(-l))|l,-l\rangle \\
&= \hbar^{2}l(l+1)|l,-l\rangle, \\
& \therefore \lambda=\hbar^{2}l(l+1).
\end{align}

$4.$ Verifique que $L_{-}|l,m\rangle = \hbar\sqrt{l(l+1)-m(m-1)}|l,m-1\rangle$.

_SOL._ Como $L_{3}(L_{-}|l,m\rangle)=\hbar(m-1)(L_{-}|l,m\rangle)$, vemos que $L_{-}|l,m\rangle\propto|l,m-1\rangle=:c|l,m-1\rangle$. Assim, tomando o produto interno desse vetor com ele mesmo,
\begin{align}
& \langle l,m|L_{-}^{\dagger}L_{-}|l,m\rangle = \langle l,m-1|c^{*}c|l,m-1\rangle, \\
& = \langle l,m|L_{+}L_{-}|l,m\rangle = |c|^{2}\langle l,m-1|l,m-1\rangle \\
& = \langle l,m|(L^{2}-L_{3}^{2}+\hbar L_{3})|l,m\rangle = \langle l,m|(\hbar^{2}l(l+1)-\hbar^{2}m^{2}+\hbar\hbar m)|l,m\rangle.
\end{align}
Desconsiderando a fase, obtemos $c=\hbar\sqrt{l(l+1)-m(m-1)}$ e $L_{-}|l,m\rangle = \hbar\sqrt{l(l+1)-m(m-1)}|l,m-1\rangle$.

$5.$ Verifique que $J^{2} = J_{a}^{2}\otimes\mathbb{I}_{b} +  \mathbb{I}_{a}\otimes J_{b}^{2} + 2\sum_{j=1}^{3}J_{j}^{a}\otimes J_{j}^{b}$.

_SOL._ Usando a definição $\vec{J} = \sum_{j=1}^{3}\hat{e}_{j}J_{j} = \sum_{k=1}^{3}\hat{e}_{j}(J_{j}^{a}\otimes\mathbb{I}_{b} + \mathbb{I}_{a}\otimes J_{j}^{b}),$ teremos

$6.$ Verifique que $[A\otimes B,A'\otimes\mathbb{I}] = [A,A']\otimes B$.

_SOL._ A verificação é uma aplicação direta de $(A\otimes B)(C\otimes D)=AC\otimes BD$:
\begin{align}
[A\otimes B,A'\otimes\mathbb{I}] & = (A\otimes B)(A'\otimes\mathbb{I}) - (A'\otimes\mathbb{I})(A\otimes B) \\
&= AA'\otimes B\mathbb{I} - A'A\otimes\mathbb{I}B = (AA'- A'A)\otimes B \\
&= [A,A']\otimes B.
\end{align}

$9.$ Verifique que $[J^{2},J_{3}^{b}]= 2i\hbar(J_{2}^{a}\otimes J_{1}^{b}-J_{1}^{a}\otimes J_{2}^{b}) \ne 0$.

\begin{align}

& = i\hbar \varepsilon_{13k}\left(J_{1}\mathbb{I}_{a}\otimes J_{k}^{b} + \mathbb{I}_{a}\otimes J_{k}^{b}J_{1} \right) + i\hbar \varepsilon_{23k}\left(J_{2}\mathbb{I}_{a}\otimes J_{k}^{b} + \mathbb{I}_{a}\otimes J_{k}^{b}J_{2} \right) \\
& = i\hbar \varepsilon_{132}\left(J_{1}\mathbb{I}_{a}\otimes J_{2}^{b} + \mathbb{I}_{a}\otimes J_{2}^{b}J_{1} \right) + i\hbar \varepsilon_{231}\left(J_{2}\mathbb{I}_{a}\otimes J_{1}^{b} + \mathbb{I}_{a}\otimes J_{1}^{b}J_{2} \right) \\
& = i\hbar\left(-J_{1}\mathbb{I}_{a}\otimes J_{2}^{b} - \mathbb{I}_{a}\otimes J_{2}^{b}J_{1} + J_{2}\mathbb{I}_{a}\otimes J_{1}^{b} + \mathbb{I}_{a}\otimes J_{1}^{b}J_{2}\right) \\
& = i\hbar\big(-(J_{1}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{1}^{b})\mathbb{I}_{a}\otimes J_{2}^{b} - \mathbb{I}_{a}\otimes J_{2}^{b}(J_{1}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{1}^{b})  \\
& \hspace{1.2cm} + (J_{2}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{2}^{b})\mathbb{I}_{a}\otimes J_{1}^{b} + \mathbb{I}_{a}\otimes J_{1}^{b}(J_{2}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{2}^{b})\big) \\
& = i\hbar\big( -J_{1}^{a}\otimes J_{2}^{b} - \mathbb{I}_{a}\otimes J_{1}^{b}J_{2}^{b} - J_{1}^{a}\otimes J_{2}^{b} - \mathbb{I}_{a}\otimes J_{2}^{b}J_{1}^{b}  \\
& \hspace{1.2cm} + J_{2}^{a}\otimes J_{1}^{b} + \mathbb{I}_{a}\otimes J_{2}^{b}J_{1}^{b}  + J_{2}^{a}\otimes J_{1}^{b} + \mathbb{I}_{a}\otimes J_{1}^{b}J_{2}^{b} \big) \\
& = 2i\hbar(J_{2}^{a}\otimes J_{1}^{b}-J_{1}^{a}\otimes J_{2}^{b}) \\
& \ne 0.                
\end{align}

$10.$ Obtenha a base $|j,m\rangle$ em termos da base $|m_{a}\rangle\otimes|m_{b}\rangle$ para $j_{a}=1$ e $j_{b}=1/2$.

_SOL._ Temos que $j_{a}=1\therefore m_{a}=-j_{a},-j_{a}+1,\cdots,j_{b}=-1,0,1$ e $j_{b}=1/2\therefore m_{b}=-1/2,1/2.$ Ademais $j=|j_{a}-j_{b}|,|j_{a}-j_{b}|+1,\cdots,j_{a}+j_{b}= 1/2,3/2.$ Tendo isso, e considerando a restrição $$m_{a}+m_{b}\ne m\Rightarrow C^{j_{a}j_{b}j}_{m_{a}m_{b}m}=0,$$ começamos com o maior valor de $j$ e maior valor de $m$: 
$$|j=3/2,m=3/2\rangle = |m_{a}=1\rangle\otimes|m_{b}=1/2\rangle.$$
Lembrando, $J_{-}|j,m\rangle=\hbar\sqrt{j(j+1)-m(m-1)}|j,m-1\rangle$ e $J_{-}=J_{-}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{-}^{b}$ e $J_{-}^{s}|m_{s}\rangle \equiv J_{-}^{s}|j_{s},m_{s}\rangle = \hbar\sqrt{j_{s}(j_{s}+1)-m_{s}(m_{s}-1)}|j_{s},m_{s}-1\rangle$ para $s=a,b.$ Assim 
\begin{align}
|j=3/2,m=1/2\rangle &= |j=3/2,m=3/2-1\rangle = \frac{1}{\hbar\sqrt{(3/2)(3/2+1)-(3/2)(3/2-1)}}J_{-}|j=3/2,m=3/2\rangle \\
& = \frac{1}{\hbar\sqrt{(3/2)(5/2)-(3/2)(1/2)}}(J_{-}^{a}\otimes\mathbb{I}_{b}+\mathbb{I}_{a}\otimes J_{-}^{b})|m_{a}=1\rangle\otimes|m_{b}=1/2\rangle \\
& = \frac{1}{\hbar\sqrt{15/4-3/4}} \big( J_{-}^{a}|m_{a}=1\rangle\otimes|m_{b}=1/2\rangle + |m_{a}=1\rangle\otimes J_{-}^{b}|m_{b}=1/2\rangle \big) \\
& = \frac{1}{\hbar\sqrt{15/4-3/4}} \big( \hbar\sqrt{1(1+1)-1(1-1)}|m_{a}=1-1\rangle\otimes|m_{b}=1/2\rangle \\
& \hspace{3.5cm} + |m_{a}=1\rangle\otimes \hbar\sqrt{(1/2)(1/2+1)-(1/2)(1/2-1)}|m_{b}=1/2-1\rangle \big) \\
& = \frac{1}{\sqrt{3}} \big( \sqrt{2}|m_{a}=0\rangle\otimes|m_{b}=1/2\rangle + |m_{a}=1\rangle\otimes \sqrt{3/4+1/4}|m_{b}=-1/2\rangle \big) \\
& = \sqrt{2/3}|m_{a}=0\rangle\otimes|m_{b}=1/2\rangle + \sqrt{1/3}|m_{a}=1\rangle\otimes |m_{b}=-1/2\rangle.
\end{align}

Usando que $J_{-}|j,m=-j\rangle = |\oslash\rangle$ teremos

& = \sqrt{1/3}|m_{a}=-1\rangle\otimes|m_{b}=1/2\rangle + \sqrt{2/3}|m_{a}=0\rangle\otimes |m_{b}=-1/2\rangle.

\end{align}

Para $j=1/2$, usamos a restrição $m_{a}+m_{b}\ne m\Rightarrow C^{j_{a}j_{b}j}_{m_{a}m_{b}m}=0$ para escrever

Por fim, aplicamos o operador de abaixamento para obter

& \hspace{1cm} +\sqrt{1/3}|m_{a}=0\rangle\otimes\sqrt{3/4+1/4}|m_{b}=-1/2\rangle \\
& = \sqrt{2/3}|m_{a}=-1\rangle\otimes|m_{b}=1/2\rangle - \sqrt{1/3}|m_{a}=0\rangle\otimes|m_{b}=-1/2\rangle,
\end{align}
que é ortogonal a $|j=3/2,m=-1/2\rangle.$

$11.$ O oscilador de Schwinger é um mapa entre a álgebra do momento angular (AMA) e a álgebra de osciladores harmônicos desacoplados. Esse mapa pode ser definido por: $J_{1}= (\hbar/2)(a_{1}^{\dagger}\otimes a_{2}+a_{1}\otimes a_{2}^{\dagger})$, $J_{2}= (\hbar/2i)(a_{1}^{\dagger}\otimes a_{2}-a_{1}\otimes a_{2}^{\dagger})$, e $J_{3} := (\hbar/2)(a_{1}^{\dagger}a_{1}\otimes\mathbb{I}_{2} - \mathbb{I}_{1}\otimes a_{2}^{\dagger}a_{2})$, com $a_{j}^{\dagger}$ e $a_{j}$  sendo, respectivamente, os operadores de criação e aniquilação do oscilador harmônico unidimensional $j=1,2$. Verifique que as componentes de $\vec{J}$, construídas dessa forma, satisfazem as relações de comutação que definem a álgebra do momento angular.

_SOL._ Usaremos 
\begin{align}
& [a_{j},a_{j}^{\dagger}]=a_{j}a_{j}^{\dagger}-a_{j}^{\dagger}a_{j}=\mathbb{I}_{j}, \\
& [a_{j},a_{j}^{\dagger}a_{j}]=a_{j}, \\
& [a_{j}^{\dagger},a_{j}^{\dagger}a_{j}]=-a_{j}^{\dagger},
\end{align}
