$1.$ Prove a desigualdade de Cauchy-Scharz, $\langle\psi|\psi\rangle\langle\phi|\phi\rangle\ge|\langle\psi|\phi\rangle|^{2}\hspace{0.2cm}\forall |\psi\rangle,|\phi\rangle\in\mathcal{H}$. Verifique que a igualdade é obtida se e somente se os vetores $|\psi\rangle$ e $|\phi\rangle$ são proporcionais.

_SOL._ Começamos usando $||\xi||^{2}\ge 0\forall|\xi\rangle\in\mathcal{H}$, e definindo $|\xi\rangle=|\psi\rangle+c|\phi\rangle$ com $c\in\mathbb{C}$. Teremos assim que
\begin{align}
& ||\xi||^{2} = \langle\xi|\xi\rangle = (|\psi\rangle+c|\phi\rangle,|\psi\rangle+c|\phi\rangle) \\
& = \langle\psi|\psi\rangle + c\langle\psi|\phi\rangle + c^{*}\langle\phi|\psi\rangle + |c|^{2}\langle\phi|\phi\rangle \ge 0.
\end{align}
Para ter algo como a DCS, podemos e vamos usar $c=-\langle\phi|\psi\rangle/\langle\phi|\phi\rangle$. Assim
\begin{align}
& \langle\psi|\psi\rangle - \frac{\langle\phi|\psi\rangle}{\langle\phi|\phi\rangle}\langle\psi|\phi\rangle - \frac{\langle\phi|\psi\rangle}{\langle\phi|\phi\rangle}^{*}\langle\phi|\psi\rangle + \left|-\frac{\langle\phi|\psi\rangle}{\langle\phi|\phi\rangle}\right|^{2}\langle\phi|\phi\rangle \ge 0, \\
& \therefore \langle\phi|\phi\rangle\langle\psi|\psi\rangle - |\langle\phi|\psi\rangle|^{2} - |\langle\phi|\psi\rangle|^{2} + |\langle\phi|\psi\rangle|^{2} \ge 0, \\
& \therefore \langle\phi|\phi\rangle\langle\psi|\psi\rangle \ge |\langle\phi|\psi\rangle|^{2}.
\end{align}

Podemos responder a segunda parte da questão usando a definição de que $||\xi||=0$ se e somente se $|\xi\rangle=|\oslash\rangle\Rightarrow |\psi\rangle=c|\phi\rangle$.

Mas podemos demostrar esse resultado escrevendo um dos vetores em termos de uma base ortonormal que contêm o outro vetor. Por exemplo, podemos escrever $|\psi\rangle=\sum_{j}c_{j}|\beta_{j}\rangle$ com $|\beta_{1}\rangle:=|\phi\rangle/||\phi||$. Com isso, assumindo igualdade na DCS, teremos
\begin{align}
& \langle\psi|\psi\rangle\langle\phi|\phi\rangle - |\langle\psi|\phi\rangle|^{2}=0, \\
& \therefore \left(\sum_{j}c_{j}|\beta_{j}\rangle,\sum_{k}c_{k}|\beta_{k}\rangle\right)\langle\phi|\phi\rangle - \left|\left(\sum_{j}c_{j}|\beta_{j}\rangle,|\phi\rangle\right)\right|^{2}=0, \\
& \therefore \sum_{j,k}c_{j}^{*}c_{k}(|\beta_{j}\rangle,|\beta_{k}\rangle)\langle\phi|\phi\rangle - \left|\sum_{j}c_{j}(|\beta_{j}\rangle,|\beta_{1}\rangle||\phi||)\right|^{2}=0, \\
& \therefore \sum_{j,k}c_{j}^{*}c_{k}\delta_{j,k}\langle\phi|\phi\rangle - \left|\sum_{j}c_{j}\delta_{j,1}||\phi||\right|^{2}=0, \\
& \therefore \sum_{j}|c_{j}|^{2}||\phi||^{2} - |c_{1}|^{2}||\phi||^{2}=0, \\
& \therefore \left(0+\sum_{j>1}|c_{j}|^{2}\right)||\phi||^{2}=0, \\
& \therefore c_{j>1}=0.
\end{align}

Por fim, se $|\psi\rangle$ e $|\phi\rangle$ são proporcionais, i.e., $|\psi\rangle=c|\phi\rangle$ com $c\in\mathbb{C}$, então $\langle\psi|\psi\rangle=(c|\phi\rangle,c|\phi\rangle)=|c|^{2}\langle\phi|\phi\rangle$ e $|\langle\psi|\phi\rangle|^{2}=|(c|\phi\rangle,|\phi\rangle)|^{2}=|c^{*}(|\phi\rangle,|\phi\rangle)|^{2}=|c|^{2}\langle\phi|\phi\rangle^{2}$, o que implica na igualdade na DCS.

$2.$  Para $|\phi\rangle=(B-\langle B\rangle\mathbb{I})|\xi\rangle$ e $B=B^{\dagger}$, verifique que $\langle\phi|\phi\rangle=Var(B)$.

_SOL._ 
\begin{align}
\langle\phi|\phi\rangle &= ((B-\langle B\rangle\mathbb{I})|\xi\rangle,(B-\langle B\rangle\mathbb{I})|\xi\rangle) = (|\xi\rangle,(B-\langle B\rangle\mathbb{I})^{\dagger}(B-\langle B\rangle\mathbb{I})|\xi\rangle) \\ 
&= (|\xi\rangle,(B^{\dagger}-\langle B\rangle^{*}\mathbb{I}^{\dagger})(B-\langle B\rangle\mathbb{I})|\xi\rangle) = (|\xi\rangle,(B-\langle B\rangle\mathbb{I})(B-\langle B\rangle\mathbb{I})|\xi\rangle) \\

&= \langle B^{2}\rangle_{\xi} - \langle B\rangle_{\xi}^{2} = Var(B)_{\xi}.
\end{align}

$3.$ Verifique que se $[A,B]=\mathbb{0}$, então $CovQ(A,B)=Cov(A,B)=Cov(B,A)$.

_SOL._ Como $[A,B]=0\Rightarrow AB=BA,$ temos que $Cov(B,A)=\langle BA\rangle-\langle B\rangle\langle A\rangle = \langle AB\rangle-\langle A\rangle\langle B\rangle = Cov(A,B).$ Assim, $CovQ(A,B)=(Cov(A,B)+Cov(B,A))/2 = Cov(A,B)=Cov(B,A).$

$4.$ Verifique que o comutador de dois operadores Hermitianos $A=A^{\dagger}$ e $B=B^{\dagger}$ é um operador anti-Hermitiano, i.e., $[A,B]^{\dagger}=-[A,B]$. Verifique que os autovalores (e portanto o valor médio) desse tipo de operador são números imaginários puros.

_SOL._ Primeiramente, temos que $[A,B]^{\dagger}=(AB-BA)^{\dagger}=(AB)^{\dagger}-(BA)^{\dagger} = B^{\dagger}A^{\dagger}-A^{\dagger}B^{\dagger} = BA-AB=(-AB-BA)=-[A,B]$. Então, para $C=[A,B]$ e $C|c\rangle=c|c\rangle$ teremos que
\begin{align}
& (C^{\dagger}|c\rangle,|c\rangle)=(-C|c\rangle,|c\rangle)=-(c|c\rangle,|c\rangle)=-c^{*}\langle c|c\rangle \\
& = (|c\rangle,C|c\rangle) = (|c\rangle,c|c\rangle) = c\langle c|c\rangle, \\
& \therefore (c+c^{*})\langle c|c\rangle = (\Re(c)+i\Im(c)+\Re(c)-i\Im(c))\langle c|c\rangle = 0, \\
& \therefore \Re(c)=0.
\end{align}

$5.$ Verifique que se o estado no qual o sistema quântico foi preparado for um autovetor do observável $A$, i.e., $|\xi\rangle=|a\rangle$ com $A|a\rangle=a|a\rangle$, então $\mathrm{Var}(A)_{\xi}=\mathrm{CovQ}(A,B)_{\xi}=\langle[A,B]\rangle_{\xi}=0$ (claro, o mesmo se verifica se $|\xi\rangle=|b\rangle$ com $B|b\rangle=b|b\rangle$).

_SOL._ Nesse caso teremos
\begin{align}
Var(A)_{a} &= \langle a|A^{2}|a\rangle-\langle a|A|a\rangle^{2} = \langle a|a^{2}|a\rangle-\langle a|a|a\rangle^{2} \\
& = a^{2}\langle a|a\rangle-a^{2}\langle a|a\rangle^{2} = a^{2}(\langle a|a\rangle-\langle a|a\rangle) = 0, \\
CovQ(A,B)_{a} &= (Cov(A,B)+Cov(B,A))/2 \\ 
&= (\langle a|AB|a\rangle - \langle a|A|a\rangle\langle a|B|a\rangle + \langle a|BA|a\rangle - \langle a|B|a\rangle\langle a|A|a\rangle)/2 \\
&= (\langle a|aB|a\rangle - \langle a|a|a\rangle\langle a|B|a\rangle + \langle a|Ba|a\rangle - \langle a|B|a\rangle\langle a|a|a\rangle)/2 \\
&= (a\langle a|B|a\rangle - a\langle a|a\rangle\langle a|B|a\rangle + a\langle a|B|a\rangle - a\langle a|B|a\rangle\langle a|a\rangle)/2 \\
&= (a\langle a|B|a\rangle - a\langle a|B|a\rangle + a\langle a|B|a\rangle - a\langle a|B|a\rangle)/2 =0, \\
\langle a|[A,B]|a\rangle &= \langle a|AB|a\rangle - \langle a|BA|a\rangle = \langle a|aB|a\rangle - \langle a|Ba|a\rangle \\
&  = a\langle a|B|a\rangle - a\langle a|B|a\rangle = 0.
\end{align}

$7.$ Considerando o grau de liberdade de spin 1/2, calcule as variâncias e o limite inferior para o produto das variâncias dos observáveis representados por $\sigma_{x}$ e $\sigma_{z}$ para o sistema no estado $|\xi\rangle = \cos(\theta/2)|z_{+}\rangle+\sin(\theta/2)|z_{-}\rangle$. Faça o gráfico de $\mathrm{Var}(\sigma_{x})$, $\mathrm{Var}(\sigma_{z})$, $\mathrm{Var}(\sigma_{x})\mathrm{Var}(\sigma_{z})$ e do limite inferior na RIHRS para esse produto em função de $\theta\in[0,\pi]$.

_SOL._ Notemos que $\langle\sigma_{x}^{2}\rangle=\langle\sigma_{z}^{2}\rangle=1.$ Ademais
\begin{align}
\langle\sigma_{x}\rangle &= \langle\xi|(\cos(\theta/2)\sigma_{x}|z_{+}\rangle+\sin(\theta/2)\sigma_{x}|z_{-}\rangle) \\
&= (\cos(\theta/2)\langle z_{+}|+\sin(\theta/2)\langle z_{-}|)(\cos(\theta/2)|z_{-}\rangle+\sin(\theta/2)|z_{+}\rangle) \\
&= 2\sin(\theta/2)\cos(\theta/2) = \sin\theta, \\
Var(\sigma_{x}) &= \langle\sigma_{x}^{2}\rangle - \langle\sigma_{x}\rangle^{2} = 1 - \sin^{2}\theta = \cos^{2}\theta, \\
\langle\sigma_{z}\rangle &= \langle\xi|(\cos(\theta/2)\sigma_{z}|z_{+}\rangle+\sin(\theta/2)\sigma_{z}|z_{-}\rangle) \\
&= (\cos(\theta/2)\langle z_{+}|+\sin(\theta/2)\langle z_{-}|)(\cos(\theta/2)|z_{+}\rangle-\sin(\theta/2)|z_{-}\rangle) \\
&= \cos^{2}(\theta/2)-\sin^{2}(\theta/2) = \cos\theta, \\
Var(\sigma_{z}) &= \langle\sigma_{z}^{2}\rangle - \langle\sigma_{z}\rangle^{2} = 1 - \cos^{2}\theta = \sin^{2}\theta, \\
Var(\sigma_{x})Var(\sigma_{z}) &= \cos^{2}\theta\sin^{2}\theta = \sin^{2}(2\theta)/4.
\end{align}

O limite inferior na RIHRS é $\mathrm{Var}(\sigma_{x})\mathrm{Var}(\sigma_{z})\ge (\mathrm{CovQ}(\sigma_{x},\sigma_{z}))^{2}+2^{-2}|\langle[\sigma_{x},\sigma_{z}]\rangle|^{2}=:L_{hrs}.$ Para calcular $L$ usamos
\begin{align}
\langle\sigma_{x}\sigma_{z}\rangle &= \langle\xi|(\cos(\theta/2)\sigma_{x}\sigma_{z}|z_{+}\rangle+\sin(\theta/2)\sigma_{x}\sigma_{z}|z_{-}\rangle) \\
&= (\cos(\theta/2)\langle z_{+}|+\sin(\theta/2)\langle z_{-}|)(\cos(\theta/2)|z_{-}\rangle-\sin(\theta/2)|z_{+}\rangle) = 0, \\
Cov(\sigma_{x},\sigma_{z}) &= \langle \sigma_{x}\sigma_{z}\rangle - \langle \sigma_{x}\rangle\langle\sigma_{z}\rangle = 0 - \sin\theta\cos\theta = -\sin(2\theta)/2, \\
\langle\sigma_{z}\sigma_{x}\rangle &= \langle\xi|(\cos(\theta/2)\sigma_{z}\sigma_{x}|z_{+}\rangle+\sin(\theta/2)\sigma_{z}\sigma_{x}|z_{-}\rangle) \\
&= (\cos(\theta/2)\langle z_{+}|+\sin(\theta/2)\langle z_{-}|)(-\cos(\theta/2)|z_{-}\rangle+\sin(\theta/2)|z_{+}\rangle) = 0, \\
\langle[\sigma_{x},\sigma_{z}]\rangle &= \langle\sigma_{x}\sigma_{z}\rangle - \langle\sigma_{z}\sigma_{x}\rangle = 0.
\end{align}
Teremos assim que
\begin{align}
L_{hrs} &= (-\sin(2\theta)/2)^{2} + |0|^{2} = \sin^{2}(2\theta)/4.
\end{align}

```python
import numpy as np
from sympy import cos, sin
from matplotlib import pyplot as plt

N = 50; dth = (np.pi-0)/N; VarX = np.zeros(N); VarZ = np.zeros(N); Lhrs = np.zeros(N); 
VarXVarZ = np.zeros(N); thv = np.zeros(N); th = -dth
for j in range(0,N):
    th += dth; thv[j] = th; VarX[j] = (cos(th))**2; VarZ[j] = (sin(th))**2
    VarXVarZ[j] = 0.25*(sin(2*th))**2; Lhrs[j] = 0.25*(sin(2*th))**2
plt.plot(thv, VarX, label=r'$Var(\sigma_{x})$'); plt.plot(thv, VarZ, label=r'$Var(\sigma_{z})$')
plt.plot(thv, VarXVarZ, label=r'$Var(\sigma_{x})Var(\sigma_{z})$'); plt.plot(thv, Lhrs, '*',label=r'$L_{hrs}$')
plt.xlabel(r'$\theta$'); axes = plt.gca(); axes.set_xlim([0,np.pi]); axes.set_ylim([-0.01,1.01])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.); plt.show()
```

$8.$ Verifique a validade geral da regra do paralelogramo $2(|||\psi\rangle||^{2}+|||\phi\rangle||^{2}) = |||\psi\rangle+|\phi\rangle||^{2} + |||\psi\rangle-|\phi\rangle||^{2}$.

_SOL._ Aplicando a definição de norma e as propriedades do produto interno, teremos
\begin{align}
& |||\psi\rangle+|\phi\rangle||^{2} + |||\psi\rangle-|\phi\rangle||^{2} = (|\psi\rangle+|\phi\rangle,|\psi\rangle+|\phi\rangle) + (|\psi\rangle-|\phi\rangle,|\psi\rangle-|\phi\rangle) \\
&= (|\psi\rangle,|\psi\rangle) + (|\psi\rangle,|\phi\rangle) +  (|\phi\rangle,|\psi\rangle) + (|\phi\rangle,|\phi\rangle) + (|\psi\rangle,|\psi\rangle) - (|\psi\rangle,|\phi\rangle) - (|\phi\rangle,|\psi\rangle) + (|\phi\rangle,|\phi\rangle) \\
&= 2(|\psi\rangle,|\psi\rangle) + 2(|\phi\rangle,|\phi\rangle) = 2(|||\psi\rangle||^{2}+|||\phi\rangle||^{2}).
\end{align}

$9.$ Calcule os limites inferiores na RIMP considerando o mesmo estado preparado e par de observáveis considerados no exercício 7. Discuta a vantagem de usarmos a RIMP no lugar da RIHRS nos casos em que $|\xi\rangle$ é autovetor de algum daqueles observáveis.

_SOL._ Neste caso, o limite inferior para $Var(\sigma_{x})+Var(\sigma_{z})$ é $L_{mp} = \max(L_{mp}^{a},L_{mp}^{b}) = \max\left(\frac{1}{2}|\langle\xi|(\sigma_{x}\pm \sigma_{z})|\xi_{\perp}\rangle|^{2},\pm i\langle[\sigma_{x},\sigma_{z}]\rangle + |\langle\xi|(\sigma_{x}\pm i\sigma_{z})|\xi_{\perp}\rangle|^{2}\right).$ Para sistemas de 2 níveis, só existe 1 estado ortogonal a $|\xi\rangle = \cos(\theta/2)|z_{+}\rangle+\sin(\theta/2)|z_{-}\rangle$, que é $|\xi_{\perp}\rangle = \sin(\theta/2)|z_{+}\rangle-\cos(\theta/2)|z_{-}\rangle.$ Com isso, teremos
\begin{align}
\langle\xi|(\sigma_{x}\pm \sigma_{z})|\xi_{\perp}\rangle &= \langle\xi|(\sigma_{x}\pm \sigma_{z})(\sin(\theta/2)|z_{+}\rangle-\cos(\theta/2)|z_{-}\rangle) \\
&= \big(\cos(\theta/2)\langle z_{+}|+\sin(\theta/2)\langle z_{-}|\big)\big(\sin(\theta/2)|z_{-}\rangle-\cos(\theta/2)|z_{+}\rangle \pm(\sin(\theta/2)|z_{+}\rangle+\cos(\theta/2)|z_{-}\rangle)\big) \\
&= -\cos^{2}(\theta/2) + \sin^{2}(\theta/2) \pm 2\sin\theta\cos\theta = -\cos\theta \pm\sin\theta, \\
L_{mp}^{a} &= \frac{1}{2}|-\cos\theta \pm\sin\theta|^{2} = 0.5\big|\cos^{2}\theta \mp2\sin\theta\cos\theta+\sin^{2}\theta\big| = (1\mp\sin\theta)/2, \\
\langle\xi|(\sigma_{x}\pm i\sigma_{z})|\xi_{\perp}\rangle &= \langle\xi|(\sigma_{x}\pm i\sigma_{z})(\sin(\theta/2)|z_{+}\rangle-\cos(\theta/2)|z_{-}\rangle) \\
&= \big(\cos(\theta/2)\langle z_{+}|+\sin(\theta/2)\langle z_{-}|\big)\big(\sin(\theta/2)|z_{-}\rangle-\cos(\theta/2)|z_{+}\rangle \pm i(\sin(\theta/2)|z_{+}\rangle+\cos(\theta/2)|z_{-}\rangle)\big) \\
&= -\cos^{2}(\theta/2) + \sin^{2}(\theta/2) \pm i2\sin\theta\cos\theta = -\cos\theta \pm i\sin\theta, \\
L_{mp}^{b} &= \pm i\langle[\sigma_{x},\sigma_{z}]\rangle+\frac{1}{2}|-\cos\theta \pm i\sin\theta|^{2} = 0+0.5\big|\cos^{2}\theta+\sin^{2}\theta\big| = 1/2.
\end{align}

```python
N = 30; dth = (np.pi-0)/N; VarX = np.zeros(N); VarZ = np.zeros(N); Lmp = np.zeros(N); 
VarXVarZ = np.zeros(N); thv = np.zeros(N); th = -dth
for j in range(0,N):
    th += dth; thv[j] = th; VarX[j] = (cos(th))**2; VarZ[j] = (sin(th))**2
    Lmp[j] = max((1-sin(th))/2,(1+sin(th))/2,1/2)
plt.plot(thv, VarX, label=r'$Var(\sigma_{x})$'); plt.plot(thv, VarZ, label=r'$Var(\sigma_{z})$')
plt.plot(thv, VarX+VarZ, label=r'$Var(\sigma_{x})+Var(\sigma_{z})$'); plt.plot(thv, Lmp, '*',label=r'$L_{mp}$')
plt.xlabel(r'$\theta$'); axes = plt.gca(); axes.set_xlim([0,np.pi]); axes.set_ylim([-0.01,1.03])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.); plt.show()
```

Vemos assim que, enquanto para $\theta=0,\pi/2,\pi$ a RIHRS não informa nada sobre a incompatibilidade entre $\sigma_{x}$ e $\sigma_{z}$, o limite inferior na RIMP é sempre positivo.

$10.$ Verifique que se um certo Hamiltoniano é independente do tempo, então seu desvio padrão é constante.

_SOL._ Se $H_{t}=H$ podemos usar $U_{t}=e^{-iHt/\hbar}\Rightarrow[H,U_{t}]=0$ e $|\psi_{t}\rangle=e^{-iHt/\hbar}|\psi_{0}\rangle$. Assim
\begin{align}
Var(H)_{\psi_{t}} &= \langle\psi_{t}|H^{2}|\psi_{t}\rangle - \langle\psi_{0}|H|\psi_{t}\rangle^{2} = \langle\psi_{0}|U_{t}^{\dagger}H^{2}U_{t}|\psi_{0}\rangle - \langle\psi_{0}|U_{t}^{\dagger}HU_{t}|\psi_{0}\rangle^{2} \\
& = \langle\psi_{0}|U_{t}^{\dagger}U_{t}H^{2}|\psi_{0}\rangle - \langle\psi_{0}|U_{t}^{\dagger}U_{t}H|\psi_{0}\rangle^{2} = \langle\psi_{0}|\mathbb{I}H^{2}|\psi_{0}\rangle - \langle\psi_{0}|\mathbb{I}H|\psi_{0}\rangle^{2} \\
& = Var(H)_{\psi_{0}} = \text{constante}.
\end{align}

$11.$ Considere o Hamiltoniano $H=-2^{-1}\hbar\omega\sigma_{z}$ e o estado inicial $|x_{+}\rangle$. Mostre que o tempo exato para que o estado evoluído seja $|x_{-}\rangle$ é igual a $\pi/\omega$ e que o limite inferior da relação $t_{\perp}\ge \pi\hbar/2\sigma_{H}$ é $\pi/\omega$.

_SOL._ Nesse caso, o operador de evolução temporal é $U_{t}=e^{-iHt/\hbar} = e^{i\omega\sigma_{z}t/2} = e^{i\omega t/2}|z_{+}\rangle\langle z_{+}| + e^{-i\omega t/2}|z_{-}\rangle\langle z_{-}|.$ Assim o estado evoluído é
\begin{align}
|\psi_{t}\rangle &= U_{t}|x_{+}\rangle = e^{i\omega t/2}|z_{+}\rangle\langle z_{+}|x_{+}\rangle + e^{-i\omega t/2}|z_{-}\rangle\langle z_{-}|x_{+}\rangle = \big(e^{i\omega t/2}|z_{+}\rangle + e^{-i\omega t/2}|z_{-}\rangle\big)/\sqrt{2} \\
& = \big(e^{i\omega t/2}(|x_{+}\rangle+|x_{+}\rangle)/\sqrt{2} + e^{-i\omega t/2}(|x_{+}\rangle-|x_{+}\rangle)/\sqrt{2}\big)/\sqrt{2} = 0.5(e^{i\omega t/2}+e^{i\omega t/2})|x_{+}\rangle + 0.5(e^{i\omega t/2}-e^{i\omega t/2})|x_{-}\rangle \\
&= \cos(\omega t/2)|x_{+}\rangle + i\sin(\omega t/2)|x_{-}\rangle.
\end{align}
O menor tempo necessário para que $|\psi_{t}\rangle \equiv |x_{-}\rangle$ é obtido quando $\sin(\omega t/2)=1,$ que é obtido quando $$\omega t/2 = \pi/2\ \therefore t=\pi/\omega.$$

Como $H$ é independente do tempo, assim também é $\sigma_{H},$ que é dado por
\begin{align}
\sigma_{H} &= \sqrt{\langle H^{2}\rangle - \langle H\rangle^{2}} = \sqrt{ \langle(\hbar^{2}\omega^{2}\mathbb{I}/4)\rangle - \langle x_{+}|(-0.5\hbar\omega\sigma_{z}|x_{+}\rangle^{2}} \\
& = \sqrt{ \hbar^{2}\omega^{2}/4 + 0.25\hbar^{2}\omega^{2}\langle x_{+}|x_{-}\rangle^{2}} = \sqrt{ \hbar^{2}\omega^{2}/4 + 0.25\hbar^{2}\omega^{2}0} \\
&= \hbar\omega/2.
\end{align}
Assim $$t_{\perp}=\frac{\pi\hbar}{2\hbar\omega/2} =\frac{\pi}{\omega}.$$
