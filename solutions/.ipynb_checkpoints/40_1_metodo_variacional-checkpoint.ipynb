$1.$ Utilize o método variacional para estimar a energia do estado fundamental do oscilador harmônico quântico unidimensional usando como função de onda teste
$$\phi(x,b) = \frac{A}{x^{2}+b^{2}}.$$

_SOL._ Começamos normalizando a função de onda
\begin{align}
& \int_{-\infty}^{+\infty}dx|\phi(x,b)|^2 = 1 \\
& \therefore\ |A|^2 \int_{-\infty}^{+\infty}\frac{dx}{(x^2 + b^2)^2}  = |A|^2 \int_{-\pi/2}^{+\pi/2}\frac{(b/\cos^2\theta) d\theta}{(b^2 / \cos^2\theta)^2}  = |A|^2 \frac{1}{b^3}\int_{-\pi/2}^{+\pi/2}\cos^2\theta d\theta = 1 \\
& \therefore\ |A|^2 \frac{1}{b^3}\int_{-\pi/2}^{+\pi/2}\Big(\frac{1}{2}+\frac{1}{2}\cos(2\theta)\Big) d\theta = |A|^2 \frac{1}{b^3}\Big(\frac{1}{2}\big(\theta\big)_{-\pi/2}^{+\pi/2} + \frac{1}{4}\big(\sin(2\theta)\big)_{-\pi/2}^{+\pi/2}\Big)  = 1 \\
& \therefore\ |A|^2 \frac{1}{b^3}\Big(\frac{1}{2}\pi + \frac{1}{4}0\Big)  = 1 \\
&\therefore\ A = \sqrt{\frac{2b^3}{\pi}}.
\end{align}
em que usamos $x=b\tan\theta\ \therefore\ dx = (b/\cos^2\theta) d\theta$ e $x^2+b^2 = b^2 / \cos^2\theta.$ Teremos assim que

\begin{align}
\langle H\rangle_\phi & = \frac{1}{2m}\langle P^2\rangle_\phi + \frac{m\omega^2}{2}\langle X^2\rangle_\phi \\
& = -\frac{\hbar^2}{2m}\frac{2b^3}{\pi}\int_{-\infty}^{+\infty}dx\frac{1}{x^2+b^2}\partial_{xx}\frac{1}{x^2+b^2} + \frac{m\omega^2}{2}\frac{2b^3}{\pi}\int_{-\infty}^{+\infty}dx\frac{x^2}{(x^2+b^2)^2} \\
& = -\frac{\hbar^2 b^3}{m\pi}\Big(8\int_{-\infty}^{+\infty}dx\frac{x^2}{(x^2+b^2)^4} - 2\int_{-\infty}^{+\infty}dx\frac{1}{(x^2+b^2)^3}\Big) + \frac{m\omega^2}{2}\frac{2b^3}{\pi}\frac{\pi}{2b} \\
\end{align}
em que usamos
\begin{align}
\int_{-\infty}^{+\infty}\frac{x^2 dx}{(x^2+b^2)^2} & = \int_{-\pi/2}^{+\pi/2}\frac{(b^2\sin^2\theta/\cos^2\theta)( b/\cos^2\theta)}{b^4/\cos^4\theta} \\
& = \frac{1}{b}\int_{-\pi/2}^{+\pi/2}\sin^2\theta d\theta \\
& = \frac{1}{b}\int_{-\pi/2}^{+\pi/2}\Big(\frac{1}{2}-\frac{1}{2}\cos 2\theta\Big) d\theta \\
& = \frac{1}{b}\frac{1}{2}\big(\pi-0\big)
\end{align}
e

\begin{align}
\partial_{xx}\big(x^2 + b^2\big)^{-1} & = \partial_{x}(-1)\big(x^2 + b^2\big)^{-2}2x \\
& = (-1)(-2)\big(x^2 + b^2\big)^{-3}2x2x + (-1)\big(x^2 + b^2\big)^{-2}2 \\
& = \frac{8x^2}{(x^2 + b^2)^3} - \frac{2}{(x^2 + b^2)^2}.
\end{align}

$2.$ Normalize a função de onda $$\phi_{1}(x) = Axe^{-bx^{2}}$$ para obter $$A = \sqrt{\frac{2^{3}b^{3/2}}{\sqrt{2\pi}}}.$$

_SOL._ Aplicamos a condição de normalização
\begin{align}
& \int_{-\infty}^{+\infty}|\phi_1(x)|^2 dx = 1 \\
& \therefore\ |A|^2\int_{-\infty}^{+\infty}x^2e^{-2bx^2}dx = 1 \\
& \therefore\ |A|^2\int_{-\infty}^{+\infty}dx\frac{\partial_b e^{-2bx^2}}{-2} = |A|^2\frac{-1}{2}\partial_b \int_{-\infty}^{+\infty}dx e^{-2bx^2} = 1 \\
& \therefore\ |A|^2\frac{-1}{2}\partial_b \sqrt{\frac{\pi}{2b}} = |A|^2\frac{-1}{2}\frac{-1}{2}b^{-3/2}\sqrt{\frac{\pi}{2}}  = |A|^2\frac{\sqrt{2\pi}}{2^3 b^{3/2}} = 1 \\
& \therefore\ A = \sqrt{\frac{2^{3}b^{3/2}}{\sqrt{2\pi}}},
\end{align}
em que assuminos $A\in\mathbb{R}$.

$3.$ Analogamente ao que fizemos no teorema variacional para o 1º estado excitado, use o método variacional para obter um limite superior para $E_{2}$, a energia do 2º estado excitado. Especifique que propriedades a função de onda deve ter.

_SOL._ Usa $|\phi\rangle$ tal que $\langle\phi|\psi_0\rangle=\langle\phi|\psi_1\rangle=0$. Assim vem que
\begin{align}
\langle H\rangle_\phi & = \sum_{n=0}^{N}E_n |\langle\phi|\psi_n\rangle|^2 \\
& = E_0 |\langle\phi|\psi_0\rangle|^2 + E_1 |\langle\phi|\psi_1\rangle|^2 + \sum_{n=2}^{N}E_n |\langle\phi|\psi_n\rangle|^2 \\
& = E_0 0 + E_1 0 + \sum_{n=2}^{N}E_n |\langle\phi|\psi_n\rangle|^2 \\
& \ge \sum_{n=2}^{N}E_2 |\langle\phi|\psi_n\rangle|^2 \\
& = E_2\big(|\langle\phi|\psi_0\rangle|^2 + E_1 |\langle\phi|\psi_1\rangle|^2 + \sum_{n=2}^{N} |\langle\phi|\psi_n\rangle|^2\big) \\
& = E_2\sum_{n=0}^{N} Pr(E_n |\phi) \\
& = E_2.
\end{align}

$4.$ Verifique que
$$\langle\phi|P^{2}X^{2}|\phi\rangle = \int dx\phi(x)^{*}\big(-i\hbar\partial_{x}\big)^{2}\big(x^{2}\phi(x)\big).$$

_SOL._ Usa $\langle x|P|\psi\rangle = -i\hbar\partial_x\langle x|\psi\rangle$ para obter
\begin{align}
\langle\phi|P^2 X^2|\phi\rangle & = \int dx\int dx'\langle\phi|P|x\rangle\langle x|P|x'\rangle\langle x'|X^2 |\phi\rangle \\
& = \int dx\int dx'\langle\phi|P|x\rangle\langle x|P|x'\rangle(x')^2\langle x'|\phi\rangle \\
& = \int dx\langle\phi|P|x\rangle\langle x|P\Big(\int dx'(x')^2\langle x'|\phi\rangle|x'\rangle\Big) \\
& = \int dx\langle\phi|P|x\rangle\frac{\hbar}{i}\partial_x\Big(\langle x|\int dx'(x')^2\langle x'|\phi\rangle|x'\rangle\Big) \\
& = \int dx\langle\phi|P|x\rangle\frac{\hbar}{i}\partial_x\Big(\int dx'(x')^2\langle x'|\phi\rangle \langle x|x'\rangle\Big) \\
& = \int dx\langle\phi|P|x\rangle\frac{\hbar}{i}\partial_x\Big(\int dx'(x')^2\langle x'|\phi\rangle \delta(x-x')\Big) \\
& = \int dx\langle\phi|P|x\rangle\frac{\hbar}{i}\partial_x\big(x^2\langle x|\phi\rangle\big) \\
& = \langle\phi|P\Big(\int dx\frac{\hbar}{i}\partial_x\big(x^2\langle x|\phi\rangle\big)|x\rangle\Big) \\
& = \int dx'\langle\phi|x'\rangle\langle x'|P\Big(\int dx\frac{\hbar}{i}\partial_x\big(x^2\langle x|\phi\rangle\big)|x\rangle\Big) \\
& = \int dx'\phi(x')^* \frac{\hbar}{i}\partial_{x'} \Big(\langle x'|\int dx\frac{\hbar}{i}\partial_x\big(x^2\langle x|\phi\rangle\big)|x\rangle\Big) \\
& = \int dx'\phi(x')^* \frac{\hbar}{i}\partial_{x'} \Big(\int dx\frac{\hbar}{i}\partial_x\big(x^2\langle x|\phi\rangle\big)\langle x'|x\rangle\Big) \\
& = \int dx'\phi(x')^* \frac{\hbar}{i}\partial_{x'}\frac{\hbar}{i}\partial_{x'}\big((x')^2\langle x'|\phi\rangle\big).
\end{align}
