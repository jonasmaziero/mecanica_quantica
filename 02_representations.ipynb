# Representações da dinâmica quântica
Assim como em Mecânica Clássica podemos descrever movimento de rotação a partir de pontos de vista diferentes mas equivalentes (e.g. partícula em movimento e referencial parado ou partícula parada e referencial em movimento), também usaremos, quando for conveniente, diferentes representações para a dinâmica quântica. 

## Representação de Schrödinger (RS)
Nessa representação o estado do sistema muda com o tempo mas os observáveis não.

Já vimos essa representação quando introduzimos os postulados da MQ. Na RS estados evoluem unitariamente ($U_{t}^{\dagger}U_{t}=U_{t}U_{t}^{\dagger}=\mathbb{I}$):
\begin{equation}
|\psi_{t}^{S}\rangle = |\psi(t)\rangle = U_t |\psi\rangle,
\end{equation}
onde $|\psi\rangle$ é o estado inicial. O operador que descreve um certo observável não muda com o tempo:
\begin{equation}
O(t) = O^{S} = O = \sum_{j}o_{j}|o_{j}\rangle\langle o_{j}|.
\end{equation}

Se o sistema foi preparado no estado $|\psi\rangle$, deixado evoluir segundo seu Hamiltoniano por um tempo $t$, e depois medimos o observável $O$, as probabilidades dos seus possíveis valores são dadas por:
\begin{equation}
Pr(o_{j}(t)) = |\langle o_{j}|\psi(t)\rangle|^{2}.
\end{equation}

### Conjunto de equações diferenciais para $|\psi_{t}\rangle$
Escrevendo o estado evoluído do sistema em termos da base de autovetores do observável de interesse:
\begin{equation}
|\psi_{t}\rangle = \sum_{j=1}^{d}c_{j}(t)|o_{j}\rangle,
\end{equation}
substituindo na equação de Schrödinger $i\hbar\partial_{t}|\psi_{t}\rangle=H_{t}|\psi_{t}\rangle$ e tomando o produto interno com $|o_{k}\rangle$ obtemos
\begin{align}
\langle o_{k}|i\hbar\partial_{t}|\psi_{t}\rangle & = \langle o_{k}|H_{t}|\psi_{t}\rangle \\
i\hbar\partial_{t}\langle o_{k}|\psi_{t}\rangle & = \langle o_{k}|H_{t}\sum_{j=1}^{d}c_{j}(t)|o_{j}\rangle \\
i\hbar\partial_{t}c_{k}(t) & = \sum_{j=1}^{d}c_{j}(t)\langle o_{k}|H_{t}|o_{j}\rangle \\
                           & = \sum_{j=1}^{d}H_{kj}c_{j}(t),
\end{align}
onde definimos os elementos de matriz do Hamiltoniano na base de autovetores do observável como: $H_{kj}:=\langle o_{k}|H_{t}|o_{j}\rangle$. Usando a notação $c_{j}(t)=c_{j}$ escreveremos essas $d$ equações como uma única equação matricial:
\begin{equation}
i\hbar 
\begin{bmatrix}
\partial_{t}c_{1} \\ \partial_{t}c_{2} \\ \vdots \\ \partial_{t}c_{d}
\end{bmatrix}
= 
\begin{bmatrix}
H_{11} & H_{12} & \cdots & H_{1d} \\
H_{21} & H_{22} & \cdots & H_{2d} \\
\vdots & \vdots & \ddots & \vdots \\
H_{d1} & H_{d2} & \cdots & H_{dd}
\end{bmatrix}
\begin{bmatrix}
c_{1} \\ c_{2} \\ \vdots \\ c_{d}
\end{bmatrix}
\text{ ou }i\hbar\partial_{t}\vec{c}=H\vec{c}.
\end{equation}
Claro, uma vez resolvido esse sistema de equações lineares (pela escolha de algum observável), teremos obtido o estado evoluído do sistema, e tudo mais que este nos fornece.

__Exercício:__ Aplicando este procedimento, obtenha $|\psi_{t}\rangle$ usando como observável o próprio Hamiltoniano (o considere independente do tempo: $H_{t}=H$). 

## Representação de Heisenberg (RH)
Nessa representação os observáveis é que mudam com o tempo, e os estados não.

O _objeto essencial_ que liga o formalismo da MQ com o experimento é o conjunto de _probabilidades_ dado pela regra de Born. Então, desde que não mudemos esses números, podemos mudar a interpretação dos objetos que aparecem no seu cálculo. Por exemplo, na equação acima se em lugar de usar $|\psi(t)\rangle=U_{t}|\psi\rangle$ usarmos $|\psi\rangle$ e no lugar de $|o_{j}\rangle$ usarmos $U_{t}^{\dagger}|o_{j}\rangle$, obteremos os mesmos valores para as probabilidades:
\begin{align}
|(U_{t}^{\dagger}| o_{j}\rangle)^{\dagger}|\psi\rangle|^{2} & = |\langle o_{j}|(U_{t}|\psi\rangle)|^{2} \\
& = |\langle o_{j}|\psi(t)\rangle)|^{2} \\ & = Pr(o_{j}(t)). 
\end{align}

Portanto, podemos considerar que o estado do sistema é sempre igual àquele preparado em $t=0$ e que são os autovetores do observável que será medido que evoluíram com o tempo, i.e.,
\begin{equation}
|\psi(t)\rangle = |\psi\rangle
\end{equation}
e
\begin{equation}
|o_{j}(t)\rangle = U_{t}^{\dagger}|o_{j}\rangle.
\end{equation}

_OBS._ Tanto na RS quanto na RH o operador de evolução temporal é obtido da equação de Schrödinger $i\hbar\partial_{t}U_{t}=H_{t}U_{t}$.

### Equação de Heisenberg
Esta equação diferencial para a evolução temporal de observáveis é útil para identificar constantes de movimento (quantidades conservadas) e é obtida como segue. Se substítuimos a equação acima na representação produto externo do observável, vem que
\begin{align}
O_{t}^{H} & = \sum_{j}o_{j}|o_{j}(t)\rangle\langle o_{j}(t)| \\ 
     & = \sum_{j}o_{j}U_{t}^{\dagger}|o_{j}\rangle\langle o_{j}|U_{t} \\
     & = U_{t}^{\dagger}\sum_{j}o_{j}|o_{j}\rangle\langle o_{j}|U_{t} \\
     & = U_{t}^{\dagger} O^{S} U_{t}=U_{t}^{\dagger} O U_{t}.
\end{align}
Derivando essa equação em relação ao tempo,
\begin{align}
i\hbar\partial_{t}O_{t}^{H} & = i\hbar\partial_{t}(U_{t}^{\dagger}O U_{t}) \\
                       & = -(-i\hbar\partial_{t}U_{t}^{\dagger})O^{S} U_{t} + 0 + U_{t}^{\dagger}O^{S}(i\hbar\partial_{t}U_{t}) \\
                       & = -U_{t}^{\dagger} H_{t}O^{S} U_{t} + U_{t}^{\dagger}O^{S} H_{t}U_{t} \\
                       & = -U_{t}^{\dagger} H_{t}\mathbb{I}O^{S} U_{t} + U_{t}^{\dagger}O^{S}\mathbb{I}H_{t}U_{t} \\
                       & = -U_{t}^{\dagger} H_{t}U_{t}U_{t}^{\dagger}O^{S} U_{t} + U_{t}^{\dagger}O^{S}U_{t}U_{t}^{\dagger}H_{t}U_{t} \\
                       & = -H_{t}^{H}O_{t}^{H} + O_{t}^{H}H_{t}^{H} \\
                       & = [O_{t}^{H},H_{t}^{H}],
\end{align}
que é a _eq. de Heisenberg._ Acima usamos usamos $H_{t}^{\dagger}=H_{t}$. 

__Exercício:__ Verifique, usando a definição de derivada, que $(i\hbar\partial_{t}U_{t})^{\dagger} = -i\hbar\partial_{t}U_{t}^{\dagger}$.

__Exercício:__ Verifique que $[U_{t},H_{t}] = 0$ se o Hamiltoniano é independente do tempo.

_OBS._ Se $[U_{t},H_{t}] = 0$, teremos $H_{t}^{H}=H_{t}$ e
\begin{equation}
i\hbar\partial_{t}O_{t}^{H} = [O_{t}^{H},H_{t}].
\end{equation}

_OBS._ Se $[O_{t}^{H},H_{t}^{H}]=0$ então $O$ é uma constante de movimento, ou quantidade conservada, i.e., a distribuição de probabilidades, valor médio e desvio padrão associados não mudam durante a evolução temporal do sistema.

#### Eq. de Heisenberg para observáveis que dependem explicitamente do tempo (na RS)
Vamos considerar observáveis que dependam explicitamente do tempo: $O_{t}^{H}=U_{t}^{\dagger}O_{t}^{s}U_{t}$. Teremos assim que
\begin{align}
i\hbar\partial_{t}O_{t}^{H} &= -(-i\hbar\partial_{t}U_{t}^{\dagger})O_{t}^{s}U_{t} + U_{t}^{\dagger}O_{t}^{s}(i\hbar\partial_{t}U_{t}) + i\hbar U_{t}^{\dagger}(\partial_{t}O_{t}^{s})U_{t} \\
&= -U_{t}^{\dagger}H_{t}O_{t}^{s}U_{t} + U_{t}^{\dagger}O_{t}^{s}H_{t}U_{t} + i\hbar U_{t}^{\dagger}(\partial_{t}O_{t}^{s})U_{t} \\
& = [O_{t}^{H},H_{t}^{H}] + i\hbar\partial_{t}^{H}O_{t}^{s},
\end{align}
onde definimos o operador derivada na representação de Heisenberg como
\begin{equation}
\partial_{t}^{H}O_{t}^{s}=U_{t}^{\dagger}(\partial_{t}O_{t}^{s})U_{t}.
\end{equation}
_OBS._ A derivada de um operador é um operador:
\begin{equation}
\partial_{t}O_{t}^{S} := \lim_{\delta t\rightarrow 0}\frac{O_{t+\delta t}^{S}-O_{t}^{S}}{\delta t}.
\end{equation}

_OBS:_ Além do Hamiltoniano, não conheço nenhum exemplo de observável que dependa explicitamente do tempo.

#### As relações de comutação são as mesmas nas RS e RI
Consideremos que a relação de comutação entre dois observáveis na representação de Schrödinger é
\begin{equation}
[A_{t}^{S},B_{t}^{S}]=C_{t}^{S}.
\end{equation}
Verifiquemos que essa relação é preservada quando mudamos para a representação de Heisenberg:
\begin{align}
[A_{t}^{H},B_{t}^{H}] & = [U_{t}^{\dagger}A_{t}^{S}U_{t},U_{t}^{\dagger}B_{t}^{S}U_{t}] \\
& = U_{t}^{\dagger}A_{t}^{S}U_{t}U_{t}^{\dagger}B_{t}^{S}U_{t} - U_{t}^{\dagger}B_{t}^{S}U_{t}U_{t}^{\dagger}A_{t}^{S}U_{t} \\
& = U_{t}^{\dagger}A_{t}^{S}B_{t}^{S}U_{t} - U_{t}^{\dagger}B_{t}^{S}A_{t}^{S}U_{t} \\
& = U_{t}^{\dagger}[A_{t}^{S},B_{t}^{S}]U_{t} =  U_{t}^{\dagger}C_{t}^{S}U_{t} \\
& = C_{t}^{H}.
\end{align}

## Representação de interação (RI)
Nessa representação tanto estados como observáveis evoluem com o tempo.

Vamos considerar sistemas físicos para os quais o Hamiltoniano pode ser escrito da seguinte forma:
\begin{equation}
H_{t} = H_{c} + V_{t},
\end{equation}
onde $H_{c} (V_{t})$ é um operador Hermitiano que não depende (depende) explicitamente do tempo. 

_OBS._ Verificaremos que na RI os estados evoluem sob $V_{t}$ e observáveis evoluem sob $H_{c}$.

Usando a parte independente do tempo do Hamiltoniano definiremos o seguinte _operador unitário:_
\begin{equation}
W_{t} = e^{-iH_{c}t/\hbar}.
\end{equation}

__Exercício:__ Verifique que $W_{t}$ é um operador unitário.

Vamos olhar agora para as probabilidades, cujos valores devem independer de representação. Denotaremos $|\psi_{s}(t)\rangle=|\psi(t)\rangle$ e $|o_{j}^{s}(t)\rangle=|o_{j}^{s}\rangle=|o_{j}\rangle$ o vetor de estado e autovetor de observável na representação de Schrödinger. Temos assim que
\begin{align}
Pr(o_{j}(t)) & = |\langle o_{j}^{s}|\psi_{s}(t)\rangle|^{2} \\
             & = |\langle o_{j}^{s}|\mathbb{I}|\psi_{s}(t)\rangle|^{2} \\
             & = |\langle o_{j}^{s}|W_{t}W^{\dagger}_{t}|\psi_{s}(t)\rangle|^{2} \\
             & = |\langle o_{j}^{I}(t)|\psi_{I}(t)\rangle|^{2},
\end{align}
com
\begin{align}
& |\psi_{I}(t)\rangle = W^{\dagger}_{t}|\psi_{s}(t)\rangle = W^{\dagger}_{t}U_{t}|\psi_{s}\rangle = W^{\dagger}_{t}U_{t}|\psi\rangle \\
& \text{ e } \\
& |o_{j}^{I}(t)\rangle = W^{\dagger}_{t}|o_{j}^{s}\rangle = W^{\dagger}_{t}|o_{j}\rangle
\end{align}
sendo, respectivamente, os vetores de estado e autovetores de observáveis evoluídos no tempo na representação de interação.

### Equação de Heisenberg na RI
Os observáveis evoluem na RI como
\begin{align}
O_{t}^{I} & = \sum_{j}o_{j}|o_{j}^{I}(t)\rangle\langle o_{j}^{I}(t)| \\
          & = \sum_{j}o_{j}W^{\dagger}_{t}|o_{j}^{s}\rangle\langle o_{j}^{s}|W_{t} \\
          & = W^{\dagger}_{t}\sum_{j}o_{j}|o_{j}^{s}\rangle\langle o_{j}^{s}|W_{t} \\
          & = W^{\dagger}_{t} O^{S} W_{t}.
\end{align}
Derivando essa equação, obtemos
\begin{align}
i\hbar\partial_{t}O_{t}^{I} & = -(-i\hbar\partial_{t}W_{t}^{\dagger}) O^{S} W_{t} + 0 + W^{\dagger}_{t} O^{S} (i\hbar\partial_{t}W_{t}).
\end{align}
Para continuar com esse cálculo, usemos a série de Taylor para $W_{t}$:
\begin{align}
i\hbar\partial_{t}W_{t} & = i\hbar\partial_{t}e^{-iH_{c}t/\hbar} \\
                        & = i\hbar\partial_{t}\sum_{j=0}^{\infty}\frac{(-iH_{c}t/\hbar)^{j}}{j!} \\
                        & = i\hbar\partial_{t}\left(\mathbb{I} + \sum_{j=1}^{\infty}\frac{(-iH_{c}/\hbar)^{j}t^{j}}{j!}\right) \\
                        & = i\hbar\left(0 + \sum_{j=1}^{\infty}\frac{(-iH_{c}/\hbar)^{j}jt^{j-1}}{j!}\right).
\end{align} 
Se trocamos de variável $k=j-1$ vem que
\begin{align}
i\hbar\partial_{t}W_{t} & = i\hbar\sum_{k=0}^{\infty}\frac{(-iH_{c}/\hbar)^{k+1}(k+1)t^{k}}{(k+1)!} \\
                        & = i\hbar(-iH_{c}/\hbar)\sum_{k=0}^{\infty}\frac{(-iH_{c}/\hbar)^{k}(k+1)t^{k}}{(k+1)(k!)} \\
                        & = H_{c}\sum_{k=0}^{\infty}\frac{(-iH_{c}t/\hbar)^{k}}{k!} \\
                        & = H_{c}e^{-iH_{c}t/\hbar} \\
                        & = H_{c}W_{t} = W_{t}H_{c}
\end{align} 
pois $[W_{t},H_{c}]=0$. Também usaremos a relação $-i\hbar\partial_{t}W_{t}^{\dagger} = W_{t}^{\dagger}H_{c}$.

__Exercício:__ Verifique que $[W_{t},H_{c}]=0$. <br>
__Exercício:__ Verifique que se dois operadores Hermitianos $A$ e $B$ comutam, também teremos que $[A,B^{\dagger}]=0$.

Com isso obteremos a _eq. de Heisenberg_:
\begin{align}
i\hbar\partial_{t}O_{t}^{I} & = -H_{c}W_{t}^{\dagger} O^{S} W_{t} + W^{\dagger}_{t} O^{S} W_{t}H_{c} \\
                            & = -H_{c}O_{t}^{I} + O_{t}^{I}H_{c} \\ 
                            & = [O_{t}^{I},H_{c}] \\
                            & = [O_{t}^{I},H_{c}^{I}],
\end{align}
pois $H_{c}^{I}=H_{c}$.

### Equação de Schrödinger na RI
Vamos derivar a equação apresentada acima para o estado evoluído na RI:
\begin{align}
i\hbar\partial_{t}|\psi_{t}^{I}\rangle & = (i\hbar\partial_{t}W^{\dagger}_{t})U_{t}|\psi_{s}\rangle + W^{\dagger}_{t}(i\hbar\partial_{t}U_{t})|\psi_{s}\rangle + 0 \\
                                      & = -H_{c}W^{\dagger}_{t}U_{t}|\psi_{s}\rangle + W^{\dagger}_{t}H_{t}U_{t}|\psi_{s}\rangle \\
                                      & = -H_{c}W^{\dagger}_{t}U_{t}|\psi_{s}\rangle + W^{\dagger}_{t}(H_c+V_{t})U_{t}|\psi_{s}\rangle \\
                                      & = W^{\dagger}_{t}V_{t}\mathbb{I}U_{t}|\psi_{s}\rangle \\
                                      & = W^{\dagger}_{t}V_{t}W_{t}W^{\dagger}_{t}U_{t}|\psi_{s}\rangle \\
                                      & = V_{t}^{I}|\psi_{t}^{I}\rangle,
\end{align}
que é a _eq. de Schrödinger_.

### Conjunto de equações diferenciais para $|\psi_{t}^{I}\rangle$
Vamos considerar que a parte independente do tempo do Hamiltoniano satisfaz a seguinte equação de autovalores e autovetores
\begin{equation}
H_{c}|E^{c}_{j}\rangle = E^{c}_{j} |E^{c}_{j}\rangle.
\end{equation}
Podemos escrever qualquer vetor de estado do sistema, e em particular $|\psi_{t}^{I}\rangle$, como combinação linear dos autovetores de $H_{c}$:
\begin{equation}
|\psi_{t}^{I}\rangle = \sum_{j=1}^{d} c_{j}(t)|E^{c}_{j}\rangle,
\end{equation}
com $c_{j}(t) = \langle E^{c}_{j}|\psi_{t}^{I}\rangle$. Como os $|E^{c}_{j}\rangle$ não dependem do tempo, vamos tomar o produto interno deles com a equação de Schödinger na RI ($i\hbar\partial_{t}|\psi_{I}(t)\rangle) = V_{t}^{I}|\psi_{t}^{I}\rangle:$
\begin{align}
\langle E^{c}_{j}|i\hbar\partial_{t}|\psi_{I}(t)\rangle & = \langle E^{c}_{j}|V_{t}^{I}|\psi_{t}^{I}\rangle \\
 i\hbar\partial_{t}\langle E^{c}_{j}|\psi_{I}(t)\rangle & = \langle E^{c}_{j}|W_{t}^{\dagger}V_{t}W_{t}\sum_{k=1}^{d} c_{k}(t)|E^{c}_{k}\rangle \\
                            i\hbar\partial_{t}c_{j}(t)  & = \sum_{k=1}^{d} c_{k}(t)\langle E^{c}_{j}|e^{iH_{c}t/\hbar}V_{t}e^{-iH_{c}t/\hbar}|E^{c}_{k}\rangle \\
                                                        & = \sum_{k=1}^{d} c_{k}(t)\langle E^{c}_{j}|e^{iE^{c}_{j}t/\hbar}V_{t}e^{-iE^{c}_{k}t/\hbar}|E^{c}_{k}\rangle \\
                                                        & = \sum_{k=1}^{d} c_{k}(t)e^{i(E^{c}_{j}-E^{c}_{k})t/\hbar}\langle E^{c}_{j}|V_{t}|E^{c}_{k}\rangle \\
                                                        & = \sum_{k=1}^{d} e^{i\omega_{jk}^{c}t}V_{jk}^{c}(t) c_{k}(t),
\end{align}
onde definimos as _frequências de Bohr_ como
\begin{equation}
\omega_{jk}^{c} = \frac{E^{c}_{j}-E^{c}_{k}}{\hbar}
\end{equation}
e os elementos de matriz da parte dependente do tempo do Hamiltoniano na base de autovetores da parte independente do tempo foram denotados por:
\begin{equation}
V_{jk}^{c}(t) = \langle E^{c}_{j}|V_{t}|E^{c}_{k}\rangle.
\end{equation}
Podemos reescrever todas essas $d$ equações em termos de uma equação matricial:
\begin{equation}
i\hbar 
\begin{bmatrix}
\partial_{t}c_{1} \\ \partial_{t}c_{2} \\ \vdots \\ \partial_{t}c_{d}
\end{bmatrix}
= 
\begin{bmatrix}
V_{11} & V_{12}e^{i\omega_{12}t} & \cdots & V_{1d}e^{i\omega_{1d}t} \\
V_{21}e^{i\omega_{21}t} & V_{22} & \cdots & V_{2d}e^{i\omega_{2d}t} \\
\vdots & \vdots & \ddots & \vdots \\
V_{d1}e^{i\omega_{d1}t} & V_{d2}e^{i\omega_{d2}t} & \cdots & V_{dd}
\end{bmatrix}
\begin{bmatrix}
c_{1} \\ c_{2} \\ \vdots \\ c_{d}
\end{bmatrix}
.
\end{equation}
Acima usamos $\omega_{jj}=0$, $c_{j}(t)=c_{j}$ e $V_{jk}(t) = V_{jk}$. Ademais, note que $\omega_{jk}=-\omega_{kj}$.

Se nosso objetivo for calcular as probabilidades dos autovalores de $H_{c}$, ao resolvermos este sistema de equações lineares teremos resolvido o problema pois
\begin{equation}
Pr(E_{j}^{c}) = |c_{j}(t)|^{2}.
\end{equation}
Por outro lado, se quisermos calcular probabilidades para um observável qualquer, precisamos obter 
\begin{equation}
|o_{j}^{I}(t)\rangle = W_{t}^{\dagger} |o_{j}^{s}\rangle = e^{iH_{c}t/\hbar}|o_{j}\rangle = \sum_{k=1}^{d}e^{iE_{k}^{c}t/\hbar}\langle E_{k}^{c}|o_{j}\rangle |E_{k}^{c}\rangle
\end{equation}
para então usá-lo junto com $|\psi_{t}^{I}\rangle$ em $Pr(o_{j}(t))=|\langle o_{j}^{I}(t)|\psi_{I}(t)\rangle)|^{2}$.
