# O operador densidade
Vimos que o significado de dizermos que um sistema quântico foi preparado em um determinado estado $|\psi\rangle$ é que sabemos que um observável foi medido e que se obteve o autovalor correspondente a este autovetor. Vamos supor que temos somente informação parcial sobre a preparação do sistema. Ou seja, considere que sabemos que o sistema foi preparado no estado $|\psi_{j}\rangle$ com probabilidade $p_{j}$, com $j=1,\cdots,n$, mas para cada quanton que nos é apresentado não sabemos qual é o valor de $j$, i.e., o índice $j$ é uma variável aleatória. Como podemos calcular probabilidades de medidas futuras de um observável $O$? Bom, para cada medida que faremos de $O$, o sistema estará definitivamente em um dos estados $\{|\psi_{j}\rangle\}_{j=1}^{n}$. Então, para o conjunto de quantons no estado $|\psi_{j}\rangle$, segue valendo a regra de Born:
\begin{equation}
Pr(o|\psi_{j}) = |\langle\psi_{j}|o\rangle|^{2}.
\end{equation}
Para obter a probabilidade para todo o ensemble $\{|\psi_{j}\rangle,p_{j}\}_{j=1}^{n}$ consideramos a média dessas probabilidades (se o índice $j$ é uma variável aleatória, a função $Pr(o|\psi_{j})$ também é):
\begin{align}
Pr(o|\{p_{j},|\psi_{j}\rangle\}) & = \sum_{j=1}^{n} Pr(o|\psi_{j})p_{j} \\
& = \sum_{j=1}^{n}|\langle o|\psi_{j}\rangle|^{2}p_{j} \\
& = \sum_{j=1}^{n} \langle o|\psi_{j}\rangle\langle\psi_{j}|o\rangle p_{j} \\
& = \langle o|\sum_{j=1}^{n} p_{j}|\psi_{j}\rangle\langle\psi_{j}|o\rangle \\
                                 & =: \langle o|\rho|o\rangle \\
                                 & \equiv Pr(o|\rho),
\end{align}
onde definimos o chamado __operador densidade__ como
\begin{equation}
\rho := \sum_{j=1}^{n} p_{j}|\psi_{j}\rangle\langle\psi_{j}|.
\end{equation}
_OBS:_ Os estados $\{|\psi_{j}\rangle\}_{j=1}^{n}$ não precisam ser mutuamente ortogonais. Ademais, $n$ pode ser menor, igual ou maior que a dimensão do espaço de estados do sistema $d=\dim\mathcal{H}$.

_OBS:_ Podem existir vários ensembles $\{|\psi_{j}\rangle,p_{j}\}_{j=1}^{n}$ diferentes que nos levam ao mesmo operador densidade. Ou seja, há arbitrariedade quando fazemos o mapa $\rho\rightarrow\{|\psi_{j}\rangle,p_{j}\}_{j=1}^{n}$, embora o mapa $\{|\psi_{j}\rangle,p_{j}\}_{j=1}^{n}\rightarrow\rho$ seja único. 

_OBS:_ No contexto do operador densidade, temos dois tipos de incerteza, a "clássica", codificada nas probabilidades dos diferentes estados do ensemble, e a "quântica", devida ao fato de que estes estados, individualmente, podem ser superposições dos autovetores do observável a ser medido.

_OBS:_ Repetibilidade continua sendo utilizada para identificar estados preparados.

A função traço é definida como 
$$\mathrm{Tr}(X)=\sum_{j}\langle b_{j}|X|b_{j}\rangle,$$
com $\{|b_{j}\rangle\}_{j=1}^{d}$ sendo qualquer base ortonormal do espaço de Hilbert sendo considerado. Sem perda de generalidade, considerando uma base $\{|b_{j}\rangle\}_{j=1}^{d}$ tal que $|b_{1}\rangle=|o\rangle$, podemos verificar que
\begin{align}
\mathrm{Tr}(\rho\Pi_{o}) & := \mathrm{Tr}(\rho|o\rangle\langle o|) \\
& = \sum_{j=1}^{d}\langle b_{j}|\rho|o\rangle\langle o|b_{j}\rangle \\
& = \sum_{j=1}^{d}\langle b_{j}|\rho|o\rangle\delta_{1j} \\
& = \langle o|\rho|o\rangle \\
& = Pr(o|\rho).
\end{align}
Além disso, como você verificará, pode-se escrever
$$\langle O\rangle_{\rho} = \mathrm{Tr}(\rho O).$$

### Exemplo
Para partículas de spin 1/2, consideremos que alguém prepara $|z_{+}\rangle$ com probabilidade $p$ ou $|x_{+}\rangle$ com probabilidade $1-p$. O operador densidade associado a esse ensemble $\{(|z_{+}\rangle,p),(|z_{-}\rangle,1-p)\}$ será:
\begin{align}
\rho & = p|z_{+}\rangle\langle z_{+}| + (1-p)|x_{+}\rangle\langle x_{+}| \\
& = p|z_{+}\rangle\langle z_{+}| + 2^{-1}(1-p)(|z_{+}\rangle+|z_{-}\rangle)(\langle z_{+}|+\langle z_{-}|) \\
     & = p|z_{+}\rangle\langle z_{+}| + 2^{-1}(1-p)(|z_{+}\rangle\langle z_{+}|+|z_{-}\rangle\langle z_{+}| + |z_{+}\rangle\langle z_{-}|+|z_{-}\rangle\langle z_{-}|) \\
     & = 2^{-1}(1+p)|z_{+}\rangle\langle z_{+}| + 2^{-1}(1-p)|z_{-}\rangle\langle z_{-}| + 2^{-1}(1-p)(|z_{-}\rangle\langle z_{+}| + |z_{+}\rangle\langle z_{-}|) \\
     & = \frac{1}{2}\begin{bmatrix} 1+p & 1-p \\ 1-p & 1-p \end{bmatrix}.
\end{align}
Acima usamos $|x_{\pm}\rangle = (|z_{+}\rangle \pm |z_{-}\rangle)/\sqrt{2}$.

Consideremos que o spin 1/2 é preparado no ensamble $\rho$ e que mediremos a componente do spin na direção $y$. Teremos as seguintes probabilidades.
Usaremos $|y_{\pm}\rangle = (|z_{+}\rangle \pm i|z_{-}\rangle)/\sqrt{2}$ e $|y_{\pm}\rangle = ((1\pm i)/2)|x_{+}\rangle + ((1\mp i)/2)|x_{-}\rangle$. Começamos com
\begin{align}
Pr(y_{+}|\rho) &= pPr(y_{+}|z_{+})+(1-p)Pr(y_{+}|x_{+}) \\
& = p|\langle y_{+}|z_{+}\rangle|^{2}+(1-p)|\langle y_{+}|x_{+}\rangle|^{2} \\
& = p|1\sqrt{2}|^{2}+(1-p)|(1+i)/2|^{2} \\
& = p/2+(1-p)((1+i)(1-i)/4 \\
& = p/2 +(1-p)/2 \\
& = 1/2.
\end{align}

Também podemos usar
\begin{align}
Pr(y_{+}|\rho) & = \langle y_{+}|\rho|y_{+}\rangle \\
& =  |y_{+}\rangle^{\dagger}|\frac{1}{2}\begin{bmatrix} 1+p & 1-p \\ 1-p & 1-p \end{bmatrix}\frac{1}{\sqrt{2}}\begin{bmatrix} 1 \\ i \end{bmatrix} \\
& = \frac{1}{\sqrt{2}}\begin{bmatrix} 1 & -i \end{bmatrix}\frac{1}{2\sqrt{2}}\begin{bmatrix} 1+p+(1-p)i \\ (1-p)(1+i) \end{bmatrix} \\
& = \frac{1}{4}(1+p+(1-p)i +(1-p)(1-i)) \\
& = 1/2.
\end{align}

Uma vez que
\begin{equation}
\Pi_{y_{+}} = |y_{+}\rangle\langle y_{+}| = \frac{1}{2} \begin{bmatrix} 1 \\ i \end{bmatrix}\begin{bmatrix} 1 & -i \end{bmatrix} = \frac{1}{2} \begin{bmatrix} 1 & -i \\ i & 1 \end{bmatrix},
\end{equation}
podemos ainda usar
\begin{align}
Pr(y_{+}|\rho) & = Tr(\rho\Pi_{y_{+}}) \\
& = Tr\left(\frac{1}{2}\begin{bmatrix} 1+p & 1-p \\ 1-p & 1-p \end{bmatrix}\frac{1}{2} \begin{bmatrix} 1 & -i \\ i & 1 \end{bmatrix}\right) \\ 
& = Tr\left(\frac{1}{4}\begin{bmatrix} (1+p)+(1-p)i & -i(1+p)+(1-p) \\ (1-p)(1+i) & (1-p)(1-i) \end{bmatrix}\right) \\
& = 1/2.
\end{align}

Por fim,
\begin{equation}
Pr(y_{-}|\rho) = 1-Pr(y_{+}|\rho)=1/2.          
\end{equation}

## Propriedades de $\rho$
Qualque operador densidade quântico válido deve possuir as seguintes propriedades:
* Traço igual a um: <br>
\begin{align}
\mathrm{Tr}(\rho) & = \sum_{j=1}^{n} p_{j}\mathrm{Tr}(|\psi_{j}\rangle\langle\psi_{j}|) \\
& = \sum_{j=1}^{n} p_{j} \\
& = 1.
\end{align}

* Hermiticidade/Positividade: <br>
Primeiramente notemos que qualquer operador densidade é um operador Hermitiano:
\begin{align}
\rho^{\dagger} & = \left(\sum_{j=1}^{n} p_{j}|\psi_{j}\rangle\langle\psi_{j}|\right)^{\dagger} \\
& = \sum_{j=1}^{n} (p_{j})^{*}(|\psi_{j}\rangle\langle\psi_{j}|)^{\dagger} \\
& = \sum_{j=1}^{n} p_{j}|\psi_{j}\rangle\langle\psi_{j}| \\
& = \rho.
\end{align}
Então, podemos diagonalizar essa matriz:
$$\rho := \sum_{j=1}^{d}r_{j}|r_{j}\rangle\langle r_{j}|.$$
Claro, podemos interpretar esse $\rho$ como representando o ensemble 
$$\rho\equiv \{|r_{j}\rangle,r_{j}\}_{j=1}^{d}.$$
Por isso os autovalores $r_{j}$ devem ser não negativos, além de $\sum_{j=1}^{d}r_{j}=1$. Garantimos positividade dos autovalores exigindo de forma geral que
\begin{equation}
\rho\ge \mathbb{O}.
\end{equation}

## Equação de von Neumann (EvN)
Considere um sistema preparado no estado 
$$\rho_{0} = \sum_{j=1}^{d} p_{j}|\psi_{j}\rangle\langle\psi_{j}| \equiv \{|\psi_{j}\rangle,p_{j}\}_{j=1}^{d}.$$
Digamos que este sistema evolui por um tempo $t$ segundo o Hamiltoniano $H_{t}$. Qual será o operador densidade evoluído desse sistema? Da ideia de ensemble sabemos que é cada um dos estados $|\psi_{j}\rangle$ que evoluirá no tempo. Ou seja, no tempo $t$ teremos o novo ensemble $\{|\psi_{j}(t)\rangle,p_{j}\}_{j=1}^{d}$ com
\begin{equation}
|\psi_{j}(t)\rangle = U_{t}|\psi_{j}\rangle.
\end{equation}
Segue assim que
\begin{align}
\rho_{t} & = \sum_{j} p_{j}|\psi_{j}(t)\rangle\langle\psi_{j}(t)| \\
& = \sum_{j} p_{j}U_{t}|\psi_{j}\rangle\langle\psi_{j}|U_{t}^{\dagger} \\
& = U_{t}\sum_{j} p_{j}|\psi_{j}\rangle\langle\psi_{j}|U_{t}^{\dagger} \\
         & = U_{t}\rho_{0} U_{t}^{\dagger}.
\end{align}

Derivando essa equação em relação ao tempo e usando a eq. de Schrödinger, $i\hbar\partial_{t}U_{t} = H_{t}U_{t}$,
obteremos a __equação de von Neumann__:
\begin{align}
i\hbar\partial_{t}\rho_{t} & = i\hbar(\partial_{t}U_{t})\rho U_{t}^{\dagger} + 0 + U_{t}\rho i\hbar\partial_{t}U_{t}^{\dagger} \\
& = H_{t}U_{t}\rho U_{t}^{\dagger} - U_{t}\rho U_{t}^{\dagger}H_{t} \\
& = H_{t}\rho_{t} - \rho_{t}H_{t} \\
                           & = [H_{t},\rho_{t}].
\end{align}

OBS. Note que aqui estamos usando a representação de Schrödinger.

### Exemplo
Consideremos o Hamiltoniano $H=-(\hbar\omega/2)\sigma_{z}$ e um estado inicial qualquer de um spin 1/2. Denotando $\rho_{jk}(t)\equiv\rho_{jk}$ teremos
\begin{align}
[H_{t},\rho_{t}] & = -\frac{\hbar\omega}{2}\left[\begin{bmatrix}1&0\\0&-1\end{bmatrix},\begin{bmatrix}\rho_{11}&\rho_{12}\\\rho_{21}&\rho_{22}\end{bmatrix}\right] \\
& = -\frac{\hbar\omega}{2}\left(\begin{bmatrix}1&0\\0&-1\end{bmatrix}\begin{bmatrix}\rho_{11}&\rho_{12}\\\rho_{21}&\rho_{22}\end{bmatrix} - \begin{bmatrix}\rho_{11}&\rho_{12}\\\rho_{21}&\rho_{22}\end{bmatrix}\begin{bmatrix}1&0\\0&-1\end{bmatrix}\right) \\
& = -\frac{\hbar\omega}{2}\left(\begin{bmatrix}\rho_{11}&\rho_{12}\\-\rho_{21}&-\rho_{22}\end{bmatrix} - \begin{bmatrix}\rho_{11}&-\rho_{12}\\ \rho_{21}&-\rho_{22}\end{bmatrix}\right) \\
& = -\frac{\hbar\omega}{2}\begin{bmatrix}0&2\rho_{12}\\-2\rho_{21}&0\end{bmatrix}
 \\ 
& = -\hbar\omega\begin{bmatrix}0&\rho_{12}\\-\rho_{21}&0\end{bmatrix}.
\end{align}
Assim
\begin{equation}
i\hbar\partial_{t}\rho_{t} = i\hbar\begin{bmatrix}\partial_{t}\rho_{11}&\partial_{t}\rho_{12}\\\partial_{t}\rho_{21}&\partial_{t}\rho_{22}\end{bmatrix}=-\hbar\omega\begin{bmatrix}0&\rho_{12}\\-\rho_{21}&0\end{bmatrix},
\end{equation}
ou
\begin{cases}
\partial_{t}\rho_{11} = 0 \therefore \rho_{11}(t)=\rho_{11}(0), \\
\partial_{t}\rho_{12} = i\omega\rho_{12} \therefore \rho_{12}(t)=\rho_{12}(0)e^{i\omega t}, \\
\rho^{\dagger}=\rho\therefore \rho_{21}(t) = (\rho_{12}(t))^{*} = (\rho_{12}(0))^{*}e^{-i\omega t} = \rho_{21}(0)e^{-i\omega t}, \\
Tr(\rho_{t})=1\therefore \rho_{22}(t) = 1-\rho_{11}(t) = 1-\rho_{11}(0)=\rho_{22}(0).
\end{cases}
Então o estado evoluído lê-se:
\begin{equation}
\rho_{t} \doteq \begin{bmatrix}\rho_{11}(0)& \rho_{12}(0)e^{i\omega t} \\ \rho_{21}(0)e^{-i\omega t} & \rho_{22}(0)\end{bmatrix}.
\end{equation}

### Exemplo: Representação de $\rho:\mathbb{C}^{2}\rightarrow\mathbb{C}^{2}$ em $\mathbb{R}^{3}$
As matrizes $\{\sigma_{j}/\sqrt{2}\}_{j=0}^{3}$ formam uma base ortonormal para $\mathbb{C}^{2x2}$, sob o produto interno de Hilbert-Schmidt $\langle A|B\rangle = \mathrm{Tr}(A^{\dagger}B)$. Por conseguinte, qualquer operador densidade pode ser escrito como
\begin{align}
\rho & = \sum_{j=0}^{3} \mathrm{Tr}\left(\rho\frac{\sigma_{j}}{\sqrt{2}}\right)\frac{\sigma_{j}}{\sqrt{2}} \\
& = \frac{1}{2}\sum_{j=0}^{3}\langle\sigma_{j}\rangle_{\rho}\sigma_{j} \\
& =: \frac{1}{2}\left(\sigma_{0} + \sum_{j=1}^{3}r_{j}\sigma_{j}\right) \\
     & =\vdots \\
     & = \frac{1}{2}\begin{bmatrix} 1+r_{3} & r_{1}-i r_{2} \\ r_{1}+i r_{2} & 1-r_{3}  \end{bmatrix},
\end{align}
em que usamos 
$$\langle\sigma_j\rangle_\rho = Tr(\sigma_j \rho)$$
e
$$\langle\sigma_{0}\rangle_{\rho}=Tr(\rho\sigma_{0})=Tr(\rho)=1.$$

_OBS:_ Os autovalores de $\rho$ são $(1\pm r)/2$, com $r = \sqrt{\sum_{j=1}^{3}r_{j}^{2}}$ e $r_{j}\in[-1,+1]$. Pela positividade do operador densidade, estados físicos estão dentro de uma bola de raio 1 em $\mathbb{R}^{3}$ centrada em (0,0,0). As componentes do ponto de $\mathbb{R}^{3}$ que representa $\rho$ são $(r_{1},r_{2},r_{3})$, que é chamado de _vetor de Bloch_.

Consideremos um spin 1/2 em um estado $\rho=\frac{1}{2}\left(\sigma_{0} + \sum_{j=1}^{3}r_{j}\sigma_{j}\right)$ e evoluindo sob o Hamiltoniano $H=-2^{-1}\hbar\omega\sigma_{3}$. Teremos assim que
\begin{align}
i\hbar\partial_{t}\rho_{t} & = [H_{t},\rho_{t}] \\
i\hbar\partial_{t}\frac{1}{2}\begin{bmatrix} 1+r_{3}(t) & r_{1}(t)-i r_{2}(t) \\ r_{1}(t)+i r_{2}(t) & 1-r_{3}(t)  \end{bmatrix} 
& = [-2^{-1}\hbar\omega\sigma_{3},2^{-1}(\sigma_{0}+r_{1}(t)\sigma_{1}+r_{2}(t)\sigma_{2}+r_{3}(t)\sigma_{3})] \\
\frac{i\hbar}{2}\begin{bmatrix} 0+\partial_{t}r_{3}(t) & \partial_{t}r_{1}(t)-i \partial_{t}r_{2}(t) \\ \partial_{t}r_{1}(t)+i \partial_{t}r_{2}(t) & 0-\partial_{t}r_{3}(t)  \end{bmatrix} 
& = -\frac{\hbar\omega}{4}[\sigma_{3},\sigma_{0}+r_{1}(t)\sigma_{1}+r_{2}(t)\sigma_{2}+r_{3}(t)\sigma_{3}] \\
                                   & = -\frac{\hbar\omega}{4}\left(\mathbb{0} + r_{1}(t)2i\epsilon_{312}\sigma_{2}+ r_{2}(t)2i\epsilon_{321}\sigma_{1}+\mathbb{0}\right) \\
                                   & = -\frac{i\hbar\omega}{2}\left(r_{1}(t)\sigma_{2} - r_{2}(t)\sigma_{1}\right) \\
                                   & = \frac{i\hbar\omega}{2}\begin{bmatrix} 0 & r_{2}(t)-ir_{1}(t) \\ r_{2}(t)+ir_{1}(t) & 0 \end{bmatrix}
\end{align}

Primeiro observe que 
$$\partial_{t}r_{3}(t) = 0 \Rightarrow r_{3}(t)=r_{3}.$$
Resta-nos resolver o seguinte sistema de eqs. diferenciais lineares de primeira ordem acopladas:
\begin{align}
\begin{cases} \partial_{t}r_{1}(t) = \omega r_{2}(t), \\ \partial_{t}r_{2}(t) = -\omega r_{1}(t).  \end{cases}
\end{align}
Do qual obtemos o sistema de eqs. diferenciais lineares de segunda ordem desacopladas:
\begin{align}
\begin{cases} \partial_{tt}r_{1}(t) = \omega \partial_{t}r_{2}(t) = -\omega^{2}r_{1}(t), \\ \partial_{tt}r_{2}(t) = -\omega \partial_{t}r_{1}(t) = -\omega^{2}r_{2}(t), \end{cases}
\end{align}
cuja solução pode ser escrita como:
\begin{align}
\begin{cases} r_{1}(t) =\langle\sigma_{1}(t)\rangle= r_{1}\cos\omega t + r_{2}\sin\omega t, \\ 
r_{2}(t) =\langle\sigma_{2}(t)\rangle= -r_{1}\sin\omega t + r_{2}\cos\omega t,\\ 
r_{3}(t)=\langle\sigma_{3}(t)\rangle=r_{3}. \end{cases}
\end{align}
Matriz densidade evoluída:
\begin{align}
\rho_{t} & = \begin{bmatrix} 1+r_{3}(t) & r_{1}(t)-ir_{2}(t) \\ r_{1}(t)+ir_{2}(t) & 1+r_{3}(t) \end{bmatrix} \\
& \vdots  \\
& = \begin{bmatrix} 1+r_{3} & (r_{1}-ir_{2})e^{i\omega t} \\ (r_{1}+ir_{2})e^{-i\omega t} & 1+r_{3} \end{bmatrix} \\
&= \begin{bmatrix} \rho_{11}(t) & \rho_{12}(t) \\ \rho_{21}(t) & \rho_{22}(t) \end{bmatrix} \\
& = \begin{bmatrix} \rho_{11}(0) & \rho_{12}(0)e^{i\omega t} \\ \rho_{21}(0)e^{-i\omega t} & \rho_{22}(0) \end{bmatrix},
\end{align}
que, claro, é a mesma solução obtida no exemplo anterior.

## Estado térmico de Gibbs
Emaranhamento do sistema com seu ambiente, que possui temperatura $T$ e é "muito grande", fará com que o estado de equilíbrio do primeiro tenha a seguinte forma https://arxiv.org/abs/quant-ph/0511225:
\begin{align}
\rho_{th} & = \frac{e^{-\beta H}}{Z} \\
& = \sum_{j=1}^{d} \frac{e^{-\beta E_{j}}}{Z}|E_{j}\rangle\langle E_{j}| \\
& = \sum_{j=1}^{d} Pr(E_{j})|E_{j}\rangle\langle E_{j}|,
\end{align}
em que 
$$\beta = 1/k_{B}T,$$
$H=\sum_{j=1}^{d}E_{j}|E_{j}\rangle\langle E_{j}|$ é o Hamiltoniano do sistema e a constante de normalização, introduzida para que $\mathrm{Tr}(\rho_{th})=1$, é a __função de partição__:
\begin{equation}
Z = \mathrm{Tr}\left(e^{-\beta H}\right) = \sum_{j=1}^{d}e^{-\beta E_{j}}.
\end{equation}

Nesse caso, podemos escrever a seguinte relação para a entropia termodinâmica (que veremos em mais detalhes a seguir):
\begin{align}
S_{th}(\rho_{th}) & := k_{B}\ln 2 S_{vN}(\rho_{th}) \\
& := - k_{B}\mathrm{Tr}(\rho_{th}\ln\rho_{th}) \\
& = -k_{B}H\left(\left\{\frac{e^{-\beta E_{j}}}{Z}\right\}\right) \\
& = - k_{B}\sum_{j=1}^{d}\frac{e^{-\beta E_{j}}}{Z}\ln\frac{e^{-\beta E_{j}}}{Z} \\
& = - k_{B}\sum_{j=1}^{d}Pr(E_{j})(\ln e^{-\beta E_{j}} - \ln Z) \\
& = - k_{B}\left(-\beta\sum_{j=1}^{d}Pr(E_{j})E_{j}-\ln Z\sum_{j=1}^{d}Pr(E_{j})\right) \\
&= k_{B}\frac{1}{k_{B}T} \langle H\rangle + k_{B}\ln Z \\
\end{align}
ou, identificando a __energia livre de Helmholtz__ 
$$F = -\beta^{-1}\ln Z,$$
teremos que
\begin{align}
F & = -k_{B}T\ln Z \\
& = \langle H\rangle_{\rho_{th}} - TS_{th}(\rho_{th}) \\
& \equiv U-TS.
\end{align}

### Exemplo
Considere um spin 1/2 com Hamiltoniano $H=-(\hbar\omega/2)\sigma_{z} = E_{+}|z_{+}\rangle\langle z_{+}|+E_{-}|z_{-}\rangle\langle z_{-}|$. Primeiro calculamos a função de partição
\begin{align}
Z & = \mathrm{Tr}(e^{-\beta H}) \\
& = \sum_{j}e^{-\beta E_{j}} \\
& = e^{-\beta E_{+}} + e^{-\beta E_{-}} \\
& = e^{-\beta (-\hbar\omega/2)} + e^{-\beta(\hbar\omega/2)} \\
  & = 2\cosh\frac{\beta\hbar\omega}{2}.
\end{align}

Estado térmico de Gibbs:
\begin{align}
\rho_{th} & = \frac{e^{-\beta H}}{Z} \\
& = \frac{e^{-\beta(-\hbar\omega/2)}|z_{+}\rangle\langle z_{+}|+e^{-\beta(\hbar\omega/2)}|z_{-}\rangle\langle z_{-}|}{Z}  \\
& = \begin{bmatrix} Z^{-1}e^{\beta\hbar\omega/2} & 0 \\ 0 & Z^{-1}e^{-\beta\hbar\omega/2} \end{bmatrix}.
\end{align}

Energia média:
\begin{align}
\langle H\rangle & = E_{+}Pr(E_{+}) + E_{-}Pr(E_{-}) \\
& = \frac{-\hbar\omega}{2}\frac{e^{\beta\hbar\omega/2}}{Z} + \frac{\hbar\omega}{2}\frac{e^{-\beta\hbar\omega/2}}{Z} \\
& \vdots \\
                 & = -\frac{\hbar\omega}{2}\tanh\frac{\beta\hbar\omega}{2}.
\end{align}
Entropia:
\begin{align}
\frac{S_{th}}{k_{B}} & = \frac{\langle H\rangle}{k_{B}T} + \ln Z \\
& = \frac{-\frac{\hbar\omega}{2}\tanh\frac{\beta\hbar\omega}{2}}{k_{B}T} + \ln\left(2\cosh\frac{\beta\hbar\omega}{2}\right) \\
                     & = -\frac{\beta\hbar\omega}{2}\tanh\frac{\beta\hbar\omega}{2} + \ln\left(2\cosh\frac{\beta\hbar\omega}{2}\right).
\end{align}

```python
def thermal(ymax):
    fig = plt.figure(figsize=(5,7)); 
    N = 150; db = (6-0)/N; bhw = -db
    bv = np.zeros(N); Pp = np.zeros(N); 
    Z = np.zeros(N); Em = np.zeros(N); Sth = np.zeros(N)
    for j in range(0,N):
        bhw += db; bv[j] = bhw; Z[j] = 2*cosh(bhw/2);
        Pp[j] = exp(bhw/2)/Z[j]
        Em[j] = -(1/2)*tanh(bhw*1/2) # E=E(\hbar\omega)
        # A entropia está em unidades de kBln2 
        Sth[j] = (-(bhw/2)*tanh(bhw/2) + ln(2*cosh(bhw/2)))/ln(2) 
    plt.xlabel(r'$\beta\hbar\omega = \hbar\omega/k_{B}T$'); 
    plt.plot(bv, Z, label=r'$Z$')
    plt.plot(bv, Pp, label=r'$Pr(E_{+})$'); 
    plt.plot(bv, Em, label=r'$\langle H\rangle/\hbar\omega$')
    plt.plot(bv, Sth, label=r'$S_{th}/k_{B}\ln2$'); 
    axes = plt.gca(); axes.set_xlim([0,6])
    axes.set_ylim([-0.51,ymax]); 
    plt.legend(bbox_to_anchor=(1.05, 1), 
               loc='upper left', borderaxespad=0.); 
    plt.show()
```

```python
interactive(thermal, ymax = (1,20))
```

# Entropia
Existe uma diferença fundamental entre a incerteza clássica associada à falta de informação sobre a preparação do sistema e a incerteza quântica associada com superposição. Esses dois tipos de incerteza podem coexistir dentro da descrição de estados via operador densidade. A seguir definiremos funções que podem ser utilizadas para quantificar essas incertezas.

## Entropia linear (EL)
Consideremos a decomposição espectral de um operador densidade, $\rho=\sum_{j=1}^{d}r_{j}|r_{j}\rangle\langle r_{j}|$. Vemos que o quadrado do operador densidade é
\begin{align}
\rho^{2} & = \sum_{j=1}^{d}r_{j}|r_{j}\rangle\langle r_{j}|\sum_{k=1}^{d}r_{k}|r_{k}\rangle\langle r_{k}| \\
& = \sum_{j,k=1}^{d}r_{j}r_{k}|r_{j}\rangle\langle r_{j}|r_{k}\rangle\langle r_{k}| \\
& = \sum_{j,k=1}^{d}r_{j}r_{k}|r_{j}\rangle\langle r_{k}|\delta_{j,k} \\
& = \sum_{j=1}^{d}r_{j}^{2}|r_{j}\rangle\langle r_{j}|.
\end{align}
Como $r_{j}\in[0,1]$ e nesse caso $r_{j}^{2}\le r_{j}$, temos que
\begin{align}
Tr(\rho^{2}) & = \sum_{j=1}^{d}r_{j}^{2}Tr(|r_{j}\rangle\langle r_{j}|) \\
& = \sum_{j=1}^{d}r_{j}^{2} \\
& \le \sum_{j=1}^{d}r_{j} \\
& = 1 \\
& = Tr(\rho).
\end{align}
No caso em que o estado do sistema é puro (não temos incerteza sobre a preparação), podemos considerar, sem perda de generalidade que 
$$\rho=|\psi_{1}\rangle\langle\psi_{1}|.$$
Assim $r_{1}^{2}=r_{1}=1$ e $r_{j\ne 1}=0$. Segue assim que nesse caso $Tr(\rho^{2})=Tr(\rho)=1$. No caso de um ensemble envolvendo estados de uma base ortonormal equiprováveis, $$\rho=\sum_{j=1}^{d}\frac{1}{d}|\phi_{j}\rangle\langle\phi_{j}|=\mathbb{I}_{d}/d,$$
temos máxima incerteza sobre a preparação e $Tr(\rho^{2}) = \sum_{j=1}^{d}(1/d)^{2} = d/d^{2} = 1/d < 1 = Tr(\rho)$. De forma geral temos
\begin{equation}
1/d \le Tr(\rho^{2}) \le 1.
\end{equation}

A __pureza__ de um operador densidade é definida com 
$$P(\rho) := \mathrm{Tr}(\rho^{2}).$$
Por sua vez, a EL é definida em termos da pureza como:
\begin{equation}
S_{L}(\rho) := 1-P(\rho)=1-Tr(\rho^{2}).
\end{equation}
Pode-se verificar que
\begin{equation}
0 = S_{L}(|\psi\rangle) \le S_{L}(\rho)\le S_{L}(\mathbb{I}/d) = (d-1)/d.
\end{equation}

## Entropia de Shannon
Para propor uma função que quantifique a informação que adquirimos ao tomar conhecimento do valor de uma variável aleatória, Shannon começou notando que a informação que obtemos ao tomar conhecimento do acontecimento de um evento $X_{j}$ é proporcional à nossa surpresa, que é inversamente proporcional a probabilidade daquele evento ocorrer:
\begin{equation}
\tilde{I}(X_{j}) = \frac{1}{p_{j}},
\end{equation}
com $p_{j}=Pr(X_{j})$. No entanto, para dois eventos independentes, $X_{j}$ e $Y_{k}$, para os quais $Pr(X_{j},Y_{k})=p_{jk}=p_{j}p_{k}=Pr(X_{j})Pr(Y_{k})$, espera-se que a informação adquirida seja a soma das informações. Mas pela medida anterior
\begin{align}
\tilde{I}(X_{j},Y_{k}) & = \frac{1}{p_{jk}} = \frac{1}{p_{j}p_{k}} \\
& \ne \frac{1}{p_{j}} + \frac{1}{p_{k}} = I(X_{j})+I(Y_{k}).
\end{align}
O uso da função logaritmo nos fornece a esperada aditividade, i.e., definindo
\begin{equation}
I(X_{j})=\log_{2}\left(\frac{1}{p_{j}}\right) = \log_{2}\left(p_{j}^{-1}\right) = - \log_{2}(p_{j}).
\end{equation}
obtemos 
\begin{align}
I(X_{j},Y_{k}) & = -\log_{2}(p_{jk}) = -\log_{2}(p_{j}p_{k}) = -\log_{2}(p_{j})-\log_{2}(p_{k}) \\
& = I(X_{j})+I(Y_{k}).
\end{align}

Para uma variável aleatória $X$ que pode assumir os valores $X_{1},\cdots,X_{n}$ com probabilidades $p_{1},\cdots,p_{n}$, a quantidade de informação que obtemos ao tomar conhecimento do valor de $X$, ou, de forma equivalente, nossa incerteza sobre $X$ antes de saber do seu valor, pode ser obtida pela _média da informação_ que obtemos ao tomar conhecimento que cada um dos seus valores aconteceu:
\begin{equation}
H(X)=\sum_{j}p_{j}I(X_{j}) = -\sum_{j}p_{j}\log_{2}(p_{j}).
\end{equation}
Essa função é chamada de __entropia de Shannon__, e sua interpretação como o número mínimo de bits de informação necessário para armazenar a informação de uma fonte que emite símbolos $\{X_{j}\}$ com probabilidade $\{p_{j}\}$ é um dos esteios da teoria matemática da informação.

OBS. Usamos $\log_{2}$ para medir entropia em bits.

OBS. Ao calcular $H$ usamos $\lim_{x\rightarrow 0} x\log x = 0$ para fazer $0\log 0 = 0$.

## Entropia de von Neumann (EvN)
Em suas investigações sobre entropia em sistemas termodinâmicos quânticos, von Neuman chegou na função
\begin{equation}
S_{vN}(\rho) = -\mathrm{Tr}(\rho\log_{2}\rho) = H(\{r_{j}\}),
\end{equation}
onde 
$$H(\{r_{j}\})=-\sum_{j}r_{j}\log_{2}r_{j}$$
é a __entropia de Shannon__ e $r_{j}$ são os autovalores de $\rho$. 

OBS. Benjamin Schumacher, em (https://doi.org/10.1103/PhysRevA.51.2738), forneceu uma interpretação operacional para a entropia de von Neuman $S_{vN}(\rho)$ como o número médio de sistemas quânticos de dois níveis (qbits) necessários para armazenar os estados da mistura em $\rho$.

Em termos de intervalos de valores, verifica-se que
\begin{equation}
0 = S_{vN}(|\psi\rangle) \le S_{vN}(\rho) \le S_{vN}(\mathbb{I}/d) = \log_{2}d.
\end{equation}

OBS. Em termodinâmica usa-se $\ln=\log_{e}$ no lugar de $\log_{2}$ e a expressão da entropia é multiplicada por $k_{B}$, a constante de Boltzmann. Obtemos com isso a __entropia termodinâmica__:
\begin{equation}
S_{th}(\rho)=k_{B}\ln(2)S_{vN}(\rho)=-k_{B}Tr(\rho\ln\rho).
\end{equation}

### Um jogo para entender a EvN
Considere os observáveis $\sigma_{x}$ e $\sigma_{z}$ de um spin 1/2 preparado em um estado $\rho$, com medidas feitas usando aparatos de Stern-Gerlach. As regras do jogo são as seguintes: Você deve escolher um único observável pra ser medido e um único autovetor desse observável. Cada vez que esse autovetor for obtido nas medidas você ganha um ponto, senão você perde um ponto. O objetivo do jogo é obter o maior saldo de pontos $N_{p}$ para um número total de medidas $N$. Vamos considerar alguns casos particulares:
* $\rho_{1}=|x_{+}\rangle\langle x_{+}|$: <br>
Nesse caso a melhor estratégia de jogo é escolher $\sigma_{x}$ e $|x_{+}\rangle$, pois nesse caso $S(|x_{+}\rangle)=0$ e sabemos que todas as medidas fornecerão o mesmo resultado, e assim $N_{p}=N_{x_{+}}-N_{x_{-}}=N-0$.

* $\rho_{2}=p_{1}|x_{+}\rangle\langle x_{+}|+p_{2}|x_{-}\rangle\langle x_{-}|$: <br>
Nesse caso, escolhendo-se $\sigma_{x}$, a melhor estratégia depende do valor de $p_{1}$. Se $p_{1}\ge 1/2$ escolhemos $|x_{+}\rangle$ obtendo $N_{p} = p_{1}N-p_{2}N=(2p_{1}-1)N$. Se $p_{1}< 1/2$ escolhe-se $|x_{-}\rangle$ obtendo $N_{p}=(2p_{2}-1)N$ pontos. A EvN desse estado é $S_{vN}(\rho_{2})=H(\{p_{1},1-p_{2}\})=-p_{1}\log_{2}(p_{1})-(1-p_{1})\log_{2}(1-p_{1})$.

* $\rho_{3}=p_{1}|x_{+}\rangle\langle x_{+}|+p_{2}|z_{+}\rangle\langle z_{+}|$: <br>
Se $p_{1}\ge 1/2$ escolhemos $\sigma_{x}$ e $|x_{+}\rangle$, e assim $N_{p}=p_{1}N+p_{2}(1/2)N-p_{2}(1/2)N=p_{1}N$. Se $p_{1}<1/2$, escolhemos $\sigma_{z}$ e $|z_{+}\rangle$, obtendo $N_{p}=p_{1}(1/2)N-p_{1}(1/2)N+p_{2}N=p_{2}N$. Nesse caso a EvN é $S_{vN}(\rho_{3})=-\lambda_{1}\log_{2}(\lambda_{1})-(1-\lambda_{1})\log_{2}(1-\lambda_{1})$ com $\lambda_{1}=(1+\sqrt{1-2p_{1}(1-p_{1})})/2$.

Veja nos gráficos abaixo a relação "inversa" que existe entre o número de pontos que fazemos e a entropia de von Neumann do estado quântico.

```python
def jogo():
    p1 = np.arange(10**-6,1,0.001); 
    Np1 = p1-p1+1; S1 = 0*p1
    Np2 = np.maximum(2*p1-1,2*(1-p1)-1); 
    S2 = -p1*np.log2(p1)-(1-p1)*np.log2(1-p1)
    Np3 = np.maximum(p1,1-p1)
    l1 = (1+np.sqrt(1-2*p1*(1-p1)))/2; 
    S3 = -l1*np.log2(l1)-(1-l1)*np.log2(1-l1)
    plt.plot(p1,Np1,'--',color='blue'); 
    plt.plot(p1,S1,color='blue')
    plt.plot(p1,Np2,'--',color='red'); 
    plt.plot(p1,S2,color='red')
    plt.plot(p1,Np3,'--',color='black'); 
    plt.plot(p1,S3,color='black')
    plt.legend((r'$N_{p}(\rho_{1})$',r'$S_{vN}(\rho_{1})$',
                r'$N_{p}(\rho_{2})$',r'$S_{vN}(\rho_{2})$',
               r'$N_{p}(\rho_{3})$',r'$S_{vN}(\rho_{3})$'),
               loc=(1.01,0.2), fontsize=13)
    plt.xlabel(r'$p_{1}$'); plt.xlim(0,1); 
    plt.ylim(-0.02,1.02)
    plt.show()
jogo()
```

## Entropia de Wigner-Yanase (EWY)
Enquanto as EL e EN quantificam incerteza clássica, a EWY quantifica incerteza quântica em relação a um observável $O$ para um sistema no estado $\rho$:
\begin{align}
S_{wy}(O|\rho) & := -\frac{1}{2}\mathrm{Tr}\left([\sqrt{\rho},O]^{2}\right) \\
&= -\frac{1}{2}\mathrm{Tr}\left((\sqrt{\rho}O-O\sqrt{\rho})^{2}\right) = -\frac{1}{2}\mathrm{Tr}\left( (\sqrt{\rho}O-O\sqrt{\rho})(\sqrt{\rho}O-O\sqrt{\rho}) \right) \\
               & = -\frac{1}{2}\mathrm{Tr}\left( \sqrt{\rho}O\sqrt{\rho}O -O\sqrt{\rho}\sqrt{\rho}O - \sqrt{\rho}OO\sqrt{\rho} +O\sqrt{\rho}O\sqrt{\rho} \right) \\
               & = -\frac{1}{2}\left( \mathrm{Tr}(\sqrt{\rho}O\sqrt{\rho}O) - \mathrm{Tr}(O\rho O) - \mathrm{Tr}(\sqrt{\rho}O^{2}\sqrt{\rho}) + \mathrm{Tr}(O\sqrt{\rho}O\sqrt{\rho}) \right) \\
               & = \mathrm{Tr}(\rho O^{2}) - \mathrm{Tr}(\sqrt{\rho}O\sqrt{\rho}O) \\
               & = \langle O^{2}\rangle_{\rho} - \mathrm{Tr}\left((O\sqrt{\rho})^{2}\right).
\end{align}

OBS. Note que se $\rho$ e $O$ comutam, a EWY é nula, mesmo que nossa incerteza clássica sobre o estado do sistema seja positiva.

OBS. Se o estado for puro, i.e., se $\rho=|\psi\rangle\langle\psi|$, a entropia de Wigner-Yanase coincide com a variância $E_{wy}(|\psi\rangle)=Var_{\psi}(O)$.

### Exemplo
_OBS:_ O comprimento do vetor de Bloch não muda por transformações unitárias. Para verificar essa informação, considera $\rho'=U\rho U^{\dagger}$. Para sistemas de 2 níveis
\begin{align}
P(\rho) & = \mathrm{Tr} \left(\frac{1}{2}\left(\sigma_{0} + \sum_{j=1}^{3}r_{j}\sigma_{j}\right)\frac{1}{2}\left(\sigma_{0} + \sum_{k=1}^{3}r_{k}\sigma_{k}\right)\right) \\
& = \frac{1}{4}\mathrm{Tr}\left(\sigma_{0} + 2\sum_{j=1}^{3}r_{j}\sigma_{j} + \sum_{j,k=1}^{3}r_{j}r_{k}\sigma_{j}\sigma_{k} \right) \\
        & = \frac{1}{4}\left(2 + 0 + \sum_{j,k=1}^{3}r_{j}r_{k}\mathrm{Tr}(\sigma_{j}\sigma_{k}) \right) = \frac{1}{4}\left(2 + \sum_{j,k=1}^{3}r_{j}r_{k}2\delta_{jk} \right) \\
        & = \frac{1}{2}\left(1 + r^{2} \right).
\end{align}
Pela invariância da função traço sob transformações unitárias (mudança de base), teremos que
\begin{align}
P(\rho') & = \frac{1}{2}\left(1 + r'^{2} \right) = \mathrm{Tr}(U\rho U^{\dagger}U\rho U^{\dagger}) = \mathrm{Tr}(U\rho\rho U^{\dagger})
          \\
          & = \mathrm{Tr}(\rho^{2}) = P(\rho).
\end{align}
Portanto $r'=r$. Ou seja, tranformações unitárias somente giram o vetor de Bloch em $\mathbb{R}^{3}$.

Vamos calcular aquelas três entropias para o estado geral de um sistema de 2 níveis mostrado acima. Da expressão para a pureza vem que
\begin{equation}
S_{L}(\rho) = \frac{1-r^{2}}{2}.
\end{equation}
Já, dos autovalores de $\rho$, a EN fica
\begin{align}
S_{vN}(\rho) & = -\frac{1+r}{2}\log_{2}\frac{1+r}{2} - \frac{1-r}{2}\log_{2}\frac{1-r}{2} \\
             & = \vdots \\
             & = 1 - 2^{-1}[(1+r)\log_{2}(1+r) + (1-r)\log_{2}(1-r)].
\end{align}


Para obter $S_{wy}$ consideraremos como observável $\sigma_{z}$. Nesse caso, como $\sigma_{z}^{2}=\sigma_{0}$, fazendo $\rho=r_{+}|r_{+}\rangle\langle r_{+}|+r_{-}|r_{-}\rangle\langle r_{-}|$ com $r_{\pm}=(1\pm r)/2$ e $|r_{+}\rangle = \cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle$ e $|r_{-}\rangle = \sin(\theta/2)|z_{+}\rangle - e^{i\phi}\cos(\theta/2)|z_{-}\rangle$. Assim, usando $\sigma_{z}|r_{+}\rangle = \cos(\theta/2)|z_{+}\rangle - e^{i\phi}\sin(\theta/2)|z_{-}\rangle$,  $\sigma_{z}|r_{-}\rangle=\sin(\theta/2)|z_{+}\rangle + e^{i\phi}\cos(\theta/2)|z_{-}\rangle$ e $\mathrm{Tr}(A|\psi\rangle\langle\psi|)=\langle\psi|A|\psi\rangle$ obteremos
\begin{align}
S_{wy}(\sigma_{z}|\rho) & = \mathrm{Tr}(\rho) - \mathrm{Tr}\left((\sigma_{z}\sqrt{\rho})^{2}\right) \\
& = 1 - \mathrm{Tr}\left((\sigma_{z}(\sqrt{r_{+}}|r_{+}\rangle\langle r_{+}|+\sqrt{r_{-}}|r_{-}\rangle\langle r_{-}|))^{2}\right) \\
                        & = 1 - \mathrm{Tr}(\sqrt{r_{+}}\sigma_{z}|r_{+}\rangle\langle r_{+}| + \sqrt{r_{-}}\sigma_{z}|r_{-}\rangle\langle r_{-}|)^{2} \\
                        & = 1 - \mathrm{Tr}\left(r_{+}\sigma_{z}|r_{+}\rangle\langle r_{+}|\sigma_{z}|r_{+}\rangle\langle r_{+}| + \sqrt{r_{+}r_{-}}\sigma_{z}|r_{+}\rangle\langle r_{+}|\sigma_{z}|r_{-}\rangle\langle r_{-}| + \sqrt{r_{+}r_{-}}\sigma_{z}|r_{-}\rangle\langle r_{-}|\sigma_{z}|r_{+}\rangle\langle r_{+}| + r_{-}\sigma_{z}|r_{-}\rangle\langle r_{-}|\sigma_{z}|r_{-}\rangle\langle r_{-}| \right) \\
                        & = 1 - r_{+}\langle r_{+}|\sigma_{z}|r_{+}\rangle^{2} - 2\sqrt{r_{+}r_{-}}|\langle r_{-}|\sigma_{z}|r_{+}\rangle|^{2} - r_{-}\langle r_{-}|\sigma_{z}|r_{-}\rangle^{2} \\
                        & = 1 - r_{+}(\cos^{2}(\theta/2)-\sin^{2}(\theta/2))^{2} - 2\sqrt{r_{+}r_{-}}|2\sin(\theta/2)\cos(\theta/2)|^{2} - r_{-}(\sin^{2}(\theta/2)-\cos^{2}(\theta/2))^{2} \\
                        & = 1 - 2^{-1}(1+r)\cos^{2}\theta - \sqrt{1-r^{2}}\sin^{2}\theta - 2^{-1}(1-r)\cos^{2}\theta  = 1 - \cos^{2}\theta - \sqrt{1-r^{2}}\sin^{2}\theta \\
                        & = (1-\sqrt{1-r^{2}})\sin^{2}\theta.
\end{align}

```python
def entropies(n):
    th = n*np.pi/8; N = 100; dr = (1.01-0)/N
    rv = np.zeros(N); P = np.zeros(N); Sl = np.zeros(N)
    Sn = np.zeros(N); 
    Swy = np.zeros(N); #th = entropies(th)
    r = -dr
    for j in range(0,N):
        r += dr; rv[j] = r; P[j] = (1+r**2)/2; Sl[j] = 1-P[j]
        Sn[j] = 1-(1/2)*((1+r)*log(1+r,2) + (1-r)*log(1-r,2)); 
        Swy[j] = (1-sqrt(1-r**2))*(sin(th))**2
    plt.xlabel(r'$r$'); plt.plot(rv, P, label=r'$P$'); 
    plt.plot(rv, Sl, label=r'$S_{L}$')
    plt.plot(rv, Sn, label=r'$S_{vN}$'); 
    plt.plot(rv, Swy, label=r'$S_{WY}(\theta=\pi/4)$')
    axes = plt.gca(); axes.set_xlim([0,1]); 
    axes.set_ylim([-0.1,1.1])
    plt.legend(bbox_to_anchor=(1.05, 1), 
               loc='upper left', borderaxespad=0.); 
    plt.show()
```

```python
interactive(entropies, n=(0,8))
```
