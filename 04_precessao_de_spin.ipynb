# Precessão de spin
Aqui veremos o 1º, e mais simples, exemplo de aplicação dos postulados da Mecânica Quântica, um sistema de dois níveis.

## Hamiltoniano
Considerando o protótipo clássico de dipolo magnético, a espira de corrente, obtemos a seguinte relação entre momento de dipolo magnético e momento angular (https://nbviewer.jupyter.org/github/jonasmaziero/mecanica_quantica_1900-1925):
\begin{equation}
\vec{\mu} = \frac{q}{2m}\vec{L}.
\end{equation}
Para descrever "_distribuições mais complicadas de corrente_", corrigimos esta expressão usando o __fator giromagnético__ $g$, que deve ser determinado experimentalmente para cada partícula em específico, e.g.
\begin{align}
& g_{e} \approx −2.0023 \text{ para o elétron,} \\
& g_{p} \approx 5.586 \text{ para o próton,} \\ 
& g_{n} \approx −3.826 \text{para o nêutron.}
\end{align}
Assim
\begin{equation}
\vec{\mu} = \frac{gq}{2m}\vec{L}.
\end{equation}

Quando nos referimos ao _momento angular intrínseco_, o spin, usamos em geral o símbolo $\vec{S}$, que por sua vez é escrito em termos de um outro operador $\vec{\Gamma}$ como
\begin{equation}
\vec{S}=\frac{\hbar}{2}\vec{\Gamma}.
\end{equation}
Juntando tudo obtemos:
\begin{equation}
\vec{\mu} = \frac{\hbar gq}{4m}\vec{\Gamma}.
\end{equation}

A _energia potencial_ de interação entre um momento e um campo magnético $\vec{B}=B\hat{n}_{B}$ é dada por:
\begin{equation}
U = -\vec{\mu}\cdot\vec{B} = -\frac{\hbar gqB}{4m}\vec{\Gamma}.\hat{n}_{B}  =: -\frac{\hbar\omega}{2}\vec{\Gamma}.\hat{n}_{B} \equiv H,
\end{equation}
desconsiderando a energia cinética. Acima definimos a _frequência de Larmor_:
\begin{equation}
\omega := \frac{g q B}{2m},
\end{equation}
que nos fornece a intensidade da interação entre $\vec{\mu}$ e $\vec{B}$. Por sua vez, a força sobre uma partícula interagindo dessa forma será
\begin{equation}
\vec{F} = -\vec{\nabla}U = \sum_{j=x,y,z}\mu_{j}\vec{\nabla}B_{j} = \sum_{j=x,y,z}\mu_{j}\left(\hat{i}\partial_{x}B_{j}+\hat{j}\partial_{y}B_{j}+\hat{k}\partial_{z}B_{j}\right).
\end{equation}

## Estados e observáveis de um spin 1/2
Em experimentos de Stern e Gerlach (ESG) (https://www.nobelprize.org/prizes/physics/1943/stern/lecture/), observa-se a quantização do spin fazendo-se partículas de spin 1/2 atravessar uma região com um gradiente de campo magnético, e observando que somente dois feixes de partículas bem separados resultam dessa interação (veja algumas animações: https://youtu.be/rg4Fnag4V-E e https://youtu.be/PH1FbkLVJU4). 

<img src='stern_gerlach.png' width='500'>

Vamos começar olhando para um detalhe que normalmente não é discutido profundamente em textos de MQ. Nesses textos, quando se explica os ESG considera-se e.g. o gradiente de campo magnético
\begin{equation}
\vec{B}= \beta z\hat{k}
\end{equation}
com $\beta$ constante. No entanto, esse campo não satisfaz a equação de Maxwell
\begin{equation}
\nabla\cdot\vec{B}=0
\end{equation}
e não é, portanto, fisicamente aceitável. Um exemplo de campo magnético que satisfaz essa equação é
\begin{equation}
\vec{B}=\beta(x\hat{i}-z\hat{k}).
\end{equation}
Mas é notável que nesse caso a força fica escrita como 
\begin{equation}
\vec{F}=-\nabla U = -\nabla (-\mu\cdot\vec{B}) = \nabla\beta(\mu_{x}x-\mu_{z}z) = \beta(\mu_{x}\hat{i} - \mu_{z}\hat{k}).
\end{equation}
Mas essa força vai separar o feixe de partículas em quatro feixes, um para cada possível par de valores das componentes do momento magnético $(\mu_{x},\mu_{z})$. Uma maneira de focar na medida de uma única componente é aplicar um campo constante muito forte na direção na qual queremos medir o momento magnético. Um exemplo de tal campo seria (veja a Ref. 2):
\begin{equation}
\vec{B}=\beta x\hat{i}+(\alpha-\beta z)\hat{k}
\end{equation}
com $\beta \ll \alpha$. Observo, no entanto, que na aplicação na interpretação de medidas em MQ, essa exclusão pela média nula de $F_{x}$ complica a interpretação quântica de colapso da função de onda. Não obstante, a seguir assumo que se possa fazer medidas de componente de momento magnético em qualquer direção que se queira.

#### Componente $z$
Consideremos o último exemplo de campo magnético, que nos fornecerá efetivamente $F_{x}=F_{y}=0$ e 
\begin{equation}
F_{z}=\mu_{z}\beta.
\end{equation}
Como as partículas são desviadas pra cima ou pra baixo, temos duas possibilidades para os estados preparados medindo o observável "componente de spin na direção $z$", que denotaremos por $\sigma_{z}$. Isso nos indica que
\begin{equation}
\text{Espaço de estados } = \mathbb{C}^{2}.
\end{equation}
Vamos denotar os estados preparados pela medida de $\sigma_{z}$ (que deve ser representado por um operador Hermitiano) por
\begin{equation}
\begin{cases}
\text{ partícula desviada pra cima em z} \Rightarrow |z_{+}\rangle \text{ foi preparado}, \\
\text{ partícula desviada pra baixo em z} \Rightarrow |z_{-}\rangle \text{ foi preparado}. 
\end{cases}
\end{equation}
Normalizaremos os autovalores para obter
\begin{equation}
\begin{cases}
\text{ partícula desviada pra cima em z} \Rightarrow z_{+} = +1, \\
\text{ partícula desviada pra baixo em z} \Rightarrow z_{-} = -1. 
\end{cases}
\end{equation}
Por conseguinte, o operador $\sigma_{z}$ e sua representação matricial ficam:
\begin{align}
\sigma_{z} & = z_{+}|z_{+}\rangle\langle z_{+}| + z_{-}|z_{-}\rangle\langle z_{-}| \\
& = |z_{+}\rangle\langle z_{+}| -|z_{-}\rangle\langle z_{-}| \\
& = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix},
\end{align}
na base ortonormal $\{|z_{+}\rangle,|z_{-}\rangle\}$.

_OBS:_ Em qualquer direção que se coloque o gradiente de campo magnético, teremos dois feixes desviados com proporções iguais. Por conseguinte, os autovalores terão a _mesma magnitude_ para medida da camponente do spin em qualquer direção.

#### Componente $x$
Determina-se experimentalmente que se prepararmos os quantons no estado $|z_{+}\rangle$ ou $|z_{-}\rangle$, através da medida de $\sigma_{z}$, e os fizermos passar por um gradiente de $\vec{B}$ na direção $x$, metade deles serão desviados pra cima e metade pra baixo. Como $\sigma_{x}$ é um operador Hermitiano, $\{|x_{\pm}\rangle\}$ é uma base ortonormal e podemos escrever
\begin{cases}
|z_{+}\rangle = c_{+}^{+}|x_{+}\rangle+c_{+}^{-}|x_{-}\rangle, \\
|z_{-}\rangle = c_{-}^{+}|x_{+}\rangle+c_{-}^{-}|x_{-}\rangle.
\end{cases}
Usando a regra de Born, temos que
\begin{equation}
Pr(x_{\pm}|z_{\pm})=|c_{\pm}^{\pm}|^{2}=1/2\therefore |c_{\pm}^{\pm}|=1/\sqrt{2}.
\end{equation}
Podemos iniciar assumindo que 
$$c_{\pm}^{\pm}\in\mathbb{R}\therefore c_{\pm}^{\pm}=\pm 1/\sqrt{2}$$ e que 
$$|z_{+}\rangle = (1/\sqrt{2})|x_{+}\rangle+(1/\sqrt{2})|x_{-}\rangle.$$
Agora, $\sigma_{z}$ é um operador Hermitiano e assim $\{|z_{\pm}\rangle\}$ é uma base ortonormal, e portanto devemos ter $\langle z_{\pm}|z_{\mp}\rangle = 0$, o que pode ser obtido fazendo 
$$|z_{-}\rangle = (1/\sqrt{2})|x_{+}\rangle-(1/\sqrt{2})|x_{-}\rangle.$$
Somando e subtraindo $\{|z_{\pm}\rangle\}$ obtemos
\begin{equation}
|x_{\pm}\rangle = \sqrt{1/2}|z_{+}\rangle \pm \sqrt{1/2}|z_{-}\rangle.
\end{equation}

Nesse caso o observável fica escrito como:
\begin{align}
\sigma_{x} &= x_{+}|x_{+}\rangle\langle x_{+}| + x_{-}|x_{-}\rangle\langle x_{-}| \\
& = (+1)|x_{+}\rangle\langle x_{+}| + (-1)|x_{-}\rangle\langle x_{-}| \\ 
& = (\sqrt{1/2}|z_{+}\rangle + \sqrt{1/2}|z_{-}\rangle)(\sqrt{1/2}\langle z_{+}| + \sqrt{1/2}\langle z_{-}|) - (\sqrt{1/2}|z_{+}\rangle - \sqrt{1/2}|z_{-}\rangle)(\sqrt{1/2}\langle z_{+}| - \sqrt{1/2}\langle z_{-}|) \\
& = (1/2)(|z_{+}\rangle\langle z_{+}| + |z_{+}\rangle\langle z_{-}| + |z_{-}\rangle\langle z_{+}| + |z_{-}\rangle\langle z_{-}|) - (1/2)(|z_{+}\rangle\langle z_{+}| - |z_{+}\rangle\langle z_{-}| - |z_{-}\rangle\langle z_{+}| + |z_{-}\rangle\langle z_{-}|) \\
& = |z_{+}\rangle\langle z_{-}| + |z_{-}\rangle\langle z_{+}| \\
& =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix},
\end{align}
na base $\{|z_{+}\rangle,|z_{-}\rangle\}$.

#### Componente $y$
Aqui começamos escrevendo
\begin{equation}
|y_{\pm}\rangle = d_{\pm}^{+}|z_{+}\rangle + d_{\pm}^{-}|z_{-}\rangle.
\end{equation}
Também é um fato empírico que se preparamos os quantons em qualquer um dos estados $\{|y_{\pm}\rangle\}$ e os fizermos passar por um gradiente de $\vec{B}$ na direção $z$, a probabilidade de desvio pra cima ou pra baixo é $1/2$. Assim, pela regra de Born, devemos ter que:
\begin{equation}
Pr(z_{\pm}|y_{\pm})=|d_{\pm}^{\pm}|^{2}=1/2 \therefore |d_{\pm}^{\pm}|=1/\sqrt{2}.
\end{equation}
Assim, se usamos 
$$d_{\pm}^{\pm}=|d_{\pm}^{\pm}|e^{i\phi_{\pm}^{\pm}}=(1/\sqrt{2})e^{i\phi_{\pm}^{\pm}},$$
teremos que
\begin{cases}
|y_{+}\rangle = (e^{i\phi_{+}^{+}}/\sqrt{2})(|z_{+}\rangle + e^{i(\phi_{+}^{-}-\phi_{+}^{+})}|z_{-}\rangle) \equiv (1/\sqrt{2})(|z_{+}\rangle + e^{i\phi_{+}}|z_{-}\rangle), \\ 
|y_{-}\rangle = (e^{i\phi_{-}^{+}}/\sqrt{2})(|z_{+}\rangle + e^{i(\phi_{-}^{-}-\phi_{-}^{+})}|z_{-}\rangle) \equiv (1/\sqrt{2})(|z_{+}\rangle + e^{i\phi_{-}}|z_{-}\rangle).
\end{cases}
Como $\sigma_{y}$ é um observável, 
\begin{align}
& \langle y_{\pm}|y_{\mp}\rangle\propto 1+e^{\pm i(\phi_{+}-\phi_{-})}=1+\cos(\phi_{+}-\phi_{-})\pm i\sin(\phi_{+}-\phi_{-})=0 \\
& \therefore\ \phi_{+}-\phi_{-}=\pi.
\end{align}

Usando e.g. $\langle x_{+}|=(1/\sqrt{2})(\langle z_{+}|+\langle z_{-}|)$ e o fato empírico $Pr(x_{+}|y_{+})=1/2$ teremos
\begin{align}
& |\langle x_{+}|y_{+}\rangle|^{2} = (1/4)|1+e^{i\phi_{+}}|^{2}=(1/2)(1+\cos\phi_{+})=1/2 \therefore \cos\phi_{+}=0 \\
& \therefore\  \phi_{+}=\pi/2=-\phi_{-}.
\end{align}
Por conseguinte
\begin{cases}
|y_{+}\rangle = (1/\sqrt{2})(|z_{+}\rangle + i|z_{-}\rangle), \\ 
|y_{-}\rangle = (1/\sqrt{2})(|z_{+}\rangle - i|z_{-}\rangle),
\end{cases}
com $i=\sqrt{-1}$. Isso nos leva à seguinte matriz que representa a componente $y$ do spin 1/2:
\begin{align}
\sigma_{y} &= y_{+}|y_{+}\rangle\langle y_{+}| + y_{-}|y_{-}\rangle\langle y_{-}| \\
& = (+1)|y_{+}\rangle\langle y_{+}| + (-1)|y_{-}\rangle\langle y_{-}| \\
& = (\sqrt{1/2}|z_{+}\rangle + i\sqrt{1/2}|z_{-}\rangle)(\sqrt{1/2}\langle z_{+}| - i\sqrt{1/2}\langle z_{-}|) - (\sqrt{1/2}|z_{+}\rangle - i\sqrt{1/2}|z_{-}\rangle)(\sqrt{1/2}\langle z_{+}| + i\sqrt{1/2}\langle z_{-}|) \\
& = (1/2)(|z_{+}\rangle\langle z_{+}| - i|z_{+}\rangle\langle z_{-}| + i|z_{-}\rangle\langle z_{+}| + |z_{-}\rangle\langle z_{-}|) - (1/2)(|z_{+}\rangle\langle z_{+}| + i|z_{+}\rangle\langle z_{-}| - i|z_{-}\rangle\langle z_{+}| + |z_{-}\rangle\langle z_{-}|) \\
& = -i|z_{+}\rangle\langle z_{-}| + i|z_{-}\rangle\langle z_{+}| \\
& =
\begin{bmatrix}
0 & -i \\
i & 0
\end{bmatrix},
\end{align}
na base $\{|z_{+}\rangle,|z_{-}\rangle\}$.

### Componente na direção $\hat{n}$
Agora, para descrever a componente do spin 1/2 em uma _direção qualquer_, consideramos o vetor de Pauli
\begin{equation}
\vec{\sigma}=\hat{i}\sigma_{x}+\hat{j}\sigma_{y}+\hat{k}\sigma_{z}
\end{equation}
e o versor naquela direção ($\theta\in[0,\pi]$, $\phi\in[0,2\pi]$),
\begin{align}
\hat{n} & = \hat{i}n_{x}+\hat{j}n_{y}+\hat{k}n_{z} \\
& = \hat{i}\sin\theta\cos\phi+\hat{j}\sin\theta\sin\phi+\hat{k}\cos\theta,
\end{align}
para escrever
\begin{align}
\hat{n}\cdot\vec{\sigma} & = \sigma_{x}\sin\theta\cos\phi + \sigma_{y}\sin\theta\sin\phi + \sigma_{z}\cos\theta \\
& =
\begin{bmatrix}
\cos\theta & e^{-i\phi}\sin\theta \\
e^{i\phi}\sin\theta & -\cos\theta
\end{bmatrix},
\end{align}
cujos autovalores são $n_{\pm}=\pm 1$ e autovetores podem ser expressos como:
\begin{cases}
|n_{+}\rangle = \cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle, \\
|n_{-}\rangle = \sin(\theta/2)|z_{+}\rangle - e^{i\phi}\cos(\theta/2)|z_{-}\rangle.
\end{cases}

OBS. As  matrizes de Pauli $\sigma_{x},\sigma_{y},\sigma_{z}$ formam uma base para matrizes $2\mathrm{x}2$ com traço nulo, como é o caso de $\hat{n}\cdot\vec{\sigma}$.

## Precessão de spin na representação de Schrödinger (RS)
Vamos considerar a evolução temporal do spin 1/2 quando este está submetido a um campo magnético constante na direção $z$, i.e., $\vec{B}=B\hat{k}$. Nesse caso o Hamiltoniano será
\begin{equation}
H = -\frac{\hbar\omega}{2}\vec{\sigma}\cdot\hat{k} = -\frac{\hbar\omega}{2}\sigma_{z} =
\begin{bmatrix}
-\frac{\hbar\omega}{2} & 0 \\
0 & \frac{\hbar\omega}{2}
\end{bmatrix}
= -\frac{\hbar\omega}{2}|z_{+}\rangle\langle z_{+}| + \frac{\hbar\omega}{2}|z_{-}\rangle\langle z_{-}|.
\end{equation}

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure(figsize=(3,2))
plt.hlines(-1,0,1,color='black')
plt.text(-0.2,-1,r'$|z_{+}\rangle$',color='black', fontsize=15)
plt.text(1.01,-1,r'$-\hbar\omega/2$',color='black', fontsize=15)
plt.hlines(1,0,1,color='black')
plt.text(-0.2,1,r'$|z_{-}\rangle$',color='black', fontsize=15)
plt.text(1.01,1,r'$\hbar\omega/2$',color='black', fontsize=15)
plt.annotate("",xy=(0.5,-1),xytext=(0.5,1),arrowprops=dict(arrowstyle="<->"))
plt.text(0.51,0,r'$\hbar\omega=h\nu$', fontsize=20); plt.axis('off')
plt.show()
```

Como o Hamiltoniano é independente do tempo, o operador de evolução temporal (que é solução de $i\hbar\partial_{t}U_{t}=H_{t}U_{t}$) pode ser escrito como
\begin{equation}
U_{t} = e^{-iHt/\hbar} = e^{i\hbar\omega\sigma_{z}t/2\hbar} = e^{i\omega\sigma_{z}t/2}.
\end{equation}

Vamos considerar que o sistema foi preparado no estado inicial 
\begin{equation}
|\psi_{0}^{S}\rangle:=|n_{+}\rangle.
\end{equation}
O estado evoluído do sistema será:
\begin{align}
|\psi_{t}^{S}\rangle & = U_{t}|\psi_{0}^{S}\rangle = U_{t}|n_{+}\rangle \\
& = U_{t}\left(\cos(\theta/2)|z_{+}\rangle + e^{i\phi}\sin(\theta/2)|z_{-}\rangle\right) \\
& = \cos(\theta/2)U_{t}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)U_{t}|z_{-}\rangle \\ 
&= \cos(\theta/2)e^{i\omega\sigma_{z}t/2}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)e^{i\omega\sigma_{z}t/2}|z_{-}\rangle \\
& = \cos(\theta/2)e^{i\omega t/2}|z_{+}\rangle + e^{i\phi}\sin(\theta/2)e^{-i\omega t/2}|z_{-}\rangle.
\end{align}

Como não mudamos as probabilidades calculadas via regra de Born pela multiplicação de um vetor de estado por uma fase global, vamos considerar
\begin{align}
|\psi_{t}^{S}\rangle & \equiv e^{-i\omega t/2}|\psi_{t}^{S}\rangle \\\
& = \cos(\theta/2)|z_{+}\rangle + e^{i(\phi -\omega t)}\sin(\theta/2)|z_{-}\rangle \\
& =: \cos(\theta_{t}/2)|z_{+}\rangle + e^{i\phi_{t}}\sin(\theta_{t}/2)|z_{-}\rangle,
\end{align}
com 
\begin{equation}
\theta_{t}=\theta=\text{constante} \text{ e } \phi_{t}=\phi-\omega t.
\end{equation}

#### Componente $z$:
Lembrando que na RS os observáveis não mudam com $t$, abaixo calculamos as probabilidades, valores médios e desvios padrões para as três componentes do spin 1/2. Será útil observar que $\sigma_{j}^{2}=\mathbb{I}$, para $j=1,2,3$, e portanto que $\langle\sigma_{j}^{2}\rangle=\langle\psi_{t}^{S}|\mathbb{I}|\psi_{t}^{S}\rangle=1$. Começando com a componente $z$ do spin:
\begin{align}
Pr(z_{+}(t)) & = |\langle z_{+}|\psi_{t}^{S}\rangle|^{2} = |\langle z_{+}|(\cos(\theta_{t}/2)|z_{+}\rangle + e^{i\phi_{t}}\sin(\theta_{t}/2)|z_{-}\rangle)|^{2} \\
& = \cos^{2}(\theta_{t}/2) = (1+\cos\theta_{t})/2, \\
Pr(z_{-}(t)) & = 1- Pr(z_{+}(t)) = (1-\cos\theta_{t})/2, \\
\langle\sigma_{z}(t)\rangle & = (-1)Pr(z_{-}(t))+(+1)Pr(z_{+}(t)) = Pr(z_{+}(t))-Pr(z_{-}(t)) = \cos\theta_{t}, \\
& \sigma_{\sigma_{z}} = \sqrt{\langle\sigma_{z}^{2}\rangle-\langle\sigma_{z}\rangle^{2}} = \sqrt{1-\cos^{2}\theta_{t}} = \sin\theta_{t}.
\end{align}
Note que a distribuição de probabilidades para $\sigma_{z}$ não muda como o tempo.

#### Componente $x$:
\begin{align}
Pr(x_{+}(t)) & = \left|\langle x_{+}|\psi_{t}^{S}\rangle\right|^{2} = \left|\langle x_{+}|(\cos(\theta_{t}/2)|z_{+}\rangle + e^{i\phi_{t}}\sin(\theta_{t}/2)|z_{-}\rangle)\right|^{2} \\
& = \left|(\cos(\theta_{t}/2) + e^{i\phi_{t}}\sin(\theta_{t}/2))/\sqrt{2}\right|^{2} \\ 
& = (\cos^{2}(\theta_{t}/2))+\sin^{2}(\theta_{t}/2)+\cos(\theta_{t}/2)\sin(\theta_{t}/2)(e^{i\phi_{t}}+e^{-i\phi_{t}}))/2 \\
& = \left(1+2\cos(\theta_{t}/2)\sin(\theta_{t}/2)\cos\phi_{t})/2 = (1+\sin\theta_{t}\cos\phi_{t}\right)/2, \\
Pr(x_{-}(t)) & = 1- Pr(x_{+}(t)) = (1-\sin\theta_{t}\cos\phi_{t})/2, \\
\langle\sigma_{x}(t)\rangle & = (-1)Pr(x_{-}(t))+(+1)Pr(x_{+}(t)) = Pr(x_{+}(t))-Pr(x_{-}(t)) = \sin\theta_{t}\cos\phi_{t}, \\
& \sigma_{\sigma_{x}} = \sqrt{\langle\sigma_{x}^{2}\rangle-\langle\sigma_{x}\rangle^{2}} = \sqrt{1-\sin^{2}\theta_{t}\cos^{2}\phi_{t}}.
\end{align}
Acima usamos $\langle x_{+}|z_{\pm}\rangle=1/\sqrt{2}$.

#### Componente $y$:
\begin{align}
Pr(y_{+}(t)) & = \left|\langle y_{+}|\psi_{t}^{S}\rangle\right|^{2} = \left|\langle y_{+}|(\cos(\theta_{t}/2)|z_{+}\rangle + e^{i\phi_{t}}\sin(\theta_{t}/2)|z_{-}\rangle)\right|^{2} \\
& = \left|(\cos(\theta_{t}/2) - ie^{i\phi_{t}}\sin(\theta_{t}/2))/\sqrt{2}\right|^{2} \\ 
& = \left(\cos^{2}(\theta_{t}/2))+\sin^{2}(\theta_{t}/2)-\cos(\theta_{t}/2)\sin(\theta_{t}/2)i(e^{i\phi_{t}}-e^{-i\phi_{t}})\right)/2 \\
& = (1+2\cos(\theta_{t}/2)\sin(\theta_{t}/2)\sin\phi_{t})/2 = (1+\sin\theta_{t}\sin\phi_{t})/2, \\
Pr(y_{-}(t)) & = 1- Pr(y_{+}(t)) = (1-\sin\theta_{t}\sin\phi_{t})/2, \\
\langle\sigma_{y}(t)\rangle & = (-1)Pr(y_{-}(t))+(+1)Pr(y_{+}(t)) = Pr(y_{+}(t))-Pr(y_{-}(t)) = \sin\theta_{t}\sin\phi_{t},  \\
& \sigma_{\sigma_{y}} = \sqrt{\langle\sigma_{y}^{2}\rangle-\langle\sigma_{y}\rangle^{2}} = \sqrt{1-\sin^{2}\theta_{t}\sin^{2}\phi_{t}}.
\end{align}
Acima usamos $\langle y_{+}|z_{+}\rangle=1/\sqrt{2}$ e $\langle y_{+}|z_{-}\rangle=-i/\sqrt{2}$.

```python
import math
from ipywidgets import interactive, interact, fixed, interact_manual, widgets
import numpy as np
```

```python
def spin_precession(th,ph):
    N = 200; dx = (2*math.pi)/N; x = np.zeros(N)
    y1 = np.zeros(N); y2 = np.zeros(N); y3 = np.zeros(N)
    y1s = np.zeros(N); y2s = np.zeros(N); y3s = np.zeros(N)
    for j in range(0,N):
        x[j] = j*dx
        y1s[j] = math.sqrt(1-math.sin(th)**2*math.cos(ph-x[j])**2)
        y1[j] = math.sin(th)*math.cos(ph-x[j])
        y2[j] = math.sin(th)*math.sin(ph-x[j]); y3[j] = math.cos(th)  
        y2s[j] = math.sqrt(1-math.sin(th)**2*math.sin(ph-x[j])**2)
        y3s[j] = math.sin(th)
    fig = plt.figure(figsize=(7,5))
    plt.plot(x,y1,color='blue'); plt.plot(x,y2,color='black')
    plt.plot(x,y3,color='red')
    plt.plot(x,y1s,color='blue',linestyle='--')
    plt.plot(x,y2s,color='black',linestyle='--')
    plt.plot(x,y3s,color='red',linestyle='--')
    plt.xlabel(r'$\omega t$')
    plt.legend((r'$\langle\sigma_{x}\rangle$',\
                r'$\langle\sigma_{y}\rangle$',\
                r'$\langle\sigma_{z}\rangle$',\
                r'$\sigma_{\sigma_{x}}$',\
                r'$\sigma_{\sigma_{y}}$',r'$\sigma_{\sigma_{z}}$')\
               ,loc=(1.01,0.2))
    plt.ylim(-1.1,1.1); plt.show()
interactive(spin_precession,th=(0,math.pi,0.05),ph=(0,4*math.pi,0.1))
```

### Representação do estado de um spin 1/2 em $\mathbb{R}^{3}$ 
Notemos que
\begin{align}
& (\langle\sigma_{x}(t)\rangle,\langle\sigma_{y}(t)\rangle,\langle\sigma_{y}(t)\rangle) = (\sin\theta_{t}\cos\phi_{t},\sin\theta_{t}\sin\phi_{t},\cos\theta_{t}) \\
& = (n_{x}(t),n_{y}(t),n_{z}(t)) = \hat{n}_{t},
\end{align}
é a direção para onde aponta o vetor em $\mathbb{R}^{3}$ que corresponde ao estado quântico $|\psi_{t}^{S}\rangle$ do spin 1/2 no instante de tempo $t$, pois se medirmos a componente do spin nessa direção obteremos $n_{+}$ com probabilidade igual a $1$, i.e.,
\begin{equation}
Pr(n_{+}(t)) = 1.
\end{equation}

Na realidade, essa evolução temporal quântica em $\mathbb{C}^{2}$ é equivalente a uma rotação em $\mathbb{R}^{3}$ em torno do eixo $z$ por um ângulo $-\omega t$:
\begin{align}
R_{z}(-\omega t)\hat{n}_{0} & = \begin{bmatrix} \cos\omega t & \sin\omega t & 0 \\ -\sin\omega t & \cos\omega t & 0 \\ 0 & 0 & 1  \end{bmatrix}\begin{bmatrix}\sin\theta\cos\phi \\ \sin\theta\sin\phi \\ \cos\theta \end{bmatrix} \\
& = \begin{bmatrix}\cos\omega t\sin\theta\cos\phi+\sin\omega t\sin\theta\sin\phi \\ -\sin\omega t\sin\theta\cos\phi+\cos\omega t\sin\theta\sin\phi \\ \cos\theta \end{bmatrix} \\ 
& = \begin{bmatrix}\sin\theta\cos(\phi-\omega t) \\ \sin\theta\sin(\phi-\omega t) \\ \cos\theta \end{bmatrix} =\begin{bmatrix} n_{x}(t) \\ n_{y}(t) \\ n_{z}(t) \end{bmatrix} = \hat{n}_{t}.
\end{align}

Abaixo está mostrado um gráfico da precessão do estado do spin 1/2 em torno da direção do campo magnético constante aplicado, tal como representado na chamada __esfera de Bloch__.

## Precessão de spin na representação de Heisenberg (RH)
Na RH, se o spin 1/2 foi preparado no estado $|n_{+}\rangle=\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle$, usaremos este mesmo vetor para calcular probabilidades em tempos posteriores. Por outro lado, evoluiremos os oberváveis. Para isso usaremos novamente: 
\begin{equation}
U_{t} = e^{i\omega\sigma_{z}t/2}= e^{i\omega t/2}|z_{+}\rangle\langle z_{+}| + e^{-i\omega t/2}|z_{-}\rangle\langle z_{-}|.
\end{equation}

#### Componente $z$:
\begin{align}
|z_{+}(t)\rangle & = U_{t}^{\dagger}|z_{+}\rangle = (e^{-i\omega t/2}|z_{+}\rangle\langle z_{+}| + e^{i\omega t/2}|z_{-}\rangle\langle z_{-}|)|z_{+}\rangle \\
                 & = e^{-i\omega t/2}|z_{+}\rangle\langle z_{+}|z_{+}\rangle + e^{i\omega t/2}|z_{-}\rangle\langle z_{-}|z_{+}\rangle \\ 
                 & = e^{-i\omega t/2}|z_{+}\rangle, \\
|z_{-}(t)\rangle & = U_{t}^{\dagger}|z_{-}\rangle = e^{i\omega t/2}|z_{-}\rangle.
\end{align}
Ou seja, estes autovetores só ganham uma fase global. Teremos assim que as probabilidades são constantes para medidas de $\sigma_{z}$:
\begin{align}
Pr(z_{+}(t)) & = |\langle z_{+}(t)|n_{+}\rangle|^{2} = |e^{i\omega t/2}\langle z_{+}|n_{+}\rangle|^{2} = |\langle z_{+}|n_{+}\rangle|^{2} = Pr(z_{+}) = (1+\cos\theta)/2, \\ 
Pr(z_{-}(t)) & = 1-Pr(z_{+}(t)) = (1-\cos\theta)/2.
\end{align}

#### Componente $x$:
\begin{align}
|x_{+}(t)\rangle & = U_{t}^{\dagger}|x_{+}\rangle = (e^{-i\omega t/2}|z_{+}\rangle\langle z_{+}| + e^{i\omega t/2}|z_{-}\rangle\langle z_{-}|)(|z_{+}\rangle+|z_{-}\rangle)/\sqrt{2} \\
& = (e^{-i\omega t/2}|z_{+}\rangle + e^{i\omega t/2}|z_{-}\rangle)/\sqrt{2} = e^{-i\omega t/2}(|z_{+}\rangle + e^{i\omega t}|z_{-}\rangle)/\sqrt{2}, \\
|x_{-}(t)\rangle & = U_{t}^{\dagger}|x_{-}\rangle = e^{-i\omega t/2}(|z_{+}\rangle - e^{i\omega t}|z_{-}\rangle)/\sqrt{2}.
\end{align}
As probabilidades serão:
\begin{align}
Pr(x_{+}(t)) & = |\langle x_{+}(t)|n_{+}\rangle|^{2} \\ 
& = \left|e^{i\omega t/2}(\langle z_{+}| + e^{-i\omega t}\langle z_{-}|)(\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle)/\sqrt{2}\right|^{2} \\
& = \left|\cos(\theta/2) + e^{i(\phi-\omega t)}\sin(\theta/2)\right|^{2}/2 \\
& = (\cos^{2}(\theta/2) + \sin(\theta/2)\cos(\theta/2)(e^{i(\phi-\omega t)}+e^{-i(\phi-\omega t)}) + \sin^{2}(\theta/2))/2 \\
& = (1+\sin\theta\cos(\phi-\omega t))/2. \\
Pr(x_{-}(t)) & = 1-Pr(x_{+}(t)) = (1-\sin\theta\cos(\phi-\omega t))/2.
\end{align}

#### Componente $y$:
\begin{align}
|y_{+}(t)\rangle & = U_{t}^{\dagger}|y_{+}\rangle = (e^{-i\omega t/2}|z_{+}\rangle\langle z_{+}| + e^{i\omega t/2}|z_{-}\rangle\langle z_{-}|)(|z_{+}\rangle+i|z_{-}\rangle)/\sqrt{2} \\
& = (e^{-i\omega t/2}|z_{+}\rangle + ie^{i\omega t/2}|z_{-}\rangle)/\sqrt{2} = e^{-i\omega t/2}(|z_{+}\rangle + ie^{i\omega t}|z_{-}\rangle)/\sqrt{2}, \\
|y_{-}(t)\rangle & = U_{t}^{\dagger}|y_{-}\rangle = e^{-i\omega t/2}(|z_{+}\rangle - ie^{i\omega t}|z_{-}\rangle)/\sqrt{2}.
\end{align}
As probabilidades serão:
\begin{align}
Pr(y_{+}(t)) & = |\langle y_{+}(t)|n_{+}\rangle|^{2} \\ 
& = \left|e^{i\omega t/2}(\langle z_{+}| - ie^{-i\omega t}\langle z_{-}|)(\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle)/\sqrt{2}\right|^{2} \\
& = \left|\cos(\theta/2) - ie^{i(\phi-\omega t)}\sin(\theta/2)\right|^{2}/2 \\ 
& = \left(\cos^{2}(\theta/2) + i\sin(\theta/2)\cos(\theta/2)(e^{-i(\phi-\omega t)}-e^{i(\phi-\omega t)})+\sin^{2}(\theta/2)\right)/2 \\
& = (1+\sin\theta\sin(\phi-\omega t))/2. \\
Pr(y_{-}(t)) & = 1-Pr(y_{+}(t)) = (1-\sin\theta\sin(\phi-\omega t))/2.
\end{align}

Obtivemos assim os mesmos resultados para as probabilidades que na RS. Os valores médios e desvios padrões são obtidos como na RS.

### Precessão de spin a partir da eq. de Heisenberg
Aqui usaremos a eq. de Heisenberg
\begin{equation}
i\hbar\partial_{t}O_{t}^{H}=[O_{t}^{H},H_{t}^{H}]+i\hbar\partial_{t}^{H}O_{t}^{S}.
\end{equation}
Nesse caso
\begin{equation}
\partial_{t}^{H}O_{t}^{S} = U_{t}^{\dagger}(\partial_{t}^{S}O_{t}^{S})U_{t}=U_{t}^{\dagger}(\mathbb{O})U_{t}=\mathbb{O},
\end{equation}
pois os observáveis considerados, $\{\sigma_{j}^{S}\}_{j=x,y,z}$, não dependem explicitamente do tempo. Usaremos também a relação de comutação geral entre matrizes de Pauli:
\begin{equation}
[\sigma_{j}^{S},\sigma_{k}^{S}]=2i\epsilon_{jkl}\sigma_{l}^{S} \therefore [\sigma_{j}^{H},\sigma_{k}^{H}]=2i\epsilon_{jkl}\sigma_{l}^{H}.
\end{equation}
Precisaremos também do Hamiltoniano
\begin{equation}
H_{t}^{H} = -\frac{\hbar\omega}{2}\sigma_{z}^{H}.
\end{equation}

Teremos assim que
\begin{align}
& i\hbar\partial_{t}\sigma_{x}^{H}  = [\sigma_{x}^{H},H^{H}] =[\sigma_{x}^{H},-\frac{\hbar\omega}{2}\sigma_{z}^{H}] = -\frac{\hbar\omega}{2}2i\epsilon_{132}\sigma_{y}^{H} \\
& \Rightarrow \partial_{t}\sigma_{x}^{H} = \omega\sigma_{y}^{H}. \\
& i\hbar\partial_{t}\sigma_{y}^{H} = [\sigma_{y}^{H},H^{H}] =[\sigma_{y}^{H},-\frac{\hbar\omega}{2}\sigma_{z}^{H}] = -\frac{\hbar\omega}{2}2i\epsilon_{231}\sigma_{y}^{H} \\
& \Rightarrow \partial_{t}\sigma_{y}^{H} = -\omega\sigma_{x}^{H}. \\
& i\hbar\partial_{t}\sigma_{z}^{H} = [\sigma_{z}^{H},H^{H}] =[\sigma_{z}^{H},-\frac{\hbar\omega}{2}\sigma_{z}^{H}] = \mathbb{0} \\
& \Rightarrow \sigma_{z}^{H} = \sigma_{z}^{S}.
\end{align}
Uma possibilidade de solução é
\begin{align}
\sigma_{x}^{H} &= \sigma_{x}^{S}\cos\omega t + \sigma_{y}^{S}\sin\omega t, \\
\sigma_{y}^{H} &= -\sigma_{x}^{S}\sin\omega t + \sigma_{y}^{S}\cos\omega t.
\end{align}
Ou, na forma matricial
\begin{equation}
\begin{bmatrix} \sigma_{x}^{H} \\ \sigma_{y}^{H} \end{bmatrix} = \begin{bmatrix} \cos\omega t & \sin\omega t \\ -\sin\omega t & \cos\omega t \end{bmatrix}\begin{bmatrix} \sigma_{x}^{S} \\ \sigma_{y}^{S} \end{bmatrix}.
\end{equation}
OBS. Note que a matriz que relaciona os "vetores de observáveis" é a matriz de rotação em $\mathbb{R}^{2}$.

Para obter os valores médios vamos usar:
\begin{cases}
\sigma_{x}|z_{+}\rangle = |z_{-}\rangle \text{, }\sigma_{x}|z_{-}\rangle = |z_{+}\rangle, \\
\sigma_{y}|z_{+}\rangle = i|z_{-}\rangle \text{, }\sigma_{y}|z_{-}\rangle = -i|z_{+}\rangle, \\
\sigma_{z}|z_{+}\rangle = |z_{+}\rangle \text{, }\sigma_{z}|z_{-}\rangle = -|z_{-}\rangle.
\end{cases}
Assim
\begin{align}
\langle\sigma_{x}\rangle_{n_{+}} & = \langle n_{+}|\sigma_{x}|n_{+}\rangle \\
& = \langle n_{+}|\sigma_{x}(\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle) \\
& = (\cos(\theta/2)\langle z_{+}|+e^{-i\phi}\sin(\theta/2)\langle z_{-}|)(\cos(\theta/2)|z_{-}\rangle+e^{i\phi}\sin(\theta/2)|z_{+}\rangle) \\
& = \sin(\theta/2)\cos(\theta/2)(e^{i\phi}+e^{-i\phi}) \\
& = 2\sin(\theta/2)\cos(\theta/2)\cos\phi \\
& = \sin\theta\cos\phi, \\
\langle\sigma_{y}\rangle_{n_{+}} & = \langle n_{+}|\sigma_{y}|n_{+}\rangle \\
& = \langle n_{+}|\sigma_{y}(\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle) \\
& = (\cos(\theta/2)\langle z_{+}|+e^{-i\phi}\sin(\theta/2)\langle z_{-}|)(i\cos(\theta/2)|z_{-}\rangle-ie^{i\phi}\sin(\theta/2)|z_{+}\rangle) \\
& = -i\sin(\theta/2)\cos(\theta/2)(e^{i\phi}-e^{-i\phi}) \\
& = -i2\sin(\theta/2)\cos(\theta/2)i\sin\phi \\
& = \sin\theta\sin\phi, \\
\langle\sigma_{z}\rangle_{n_{+}} & = \langle n_{+}|\sigma_{z}|n_{+}\rangle \\
& = \langle n_{+}|\sigma_{z}(\cos(\theta/2)|z_{+}\rangle+e^{i\phi}\sin(\theta/2)|z_{-}\rangle) \\
& = (\cos(\theta/2)\langle z_{+}|+e^{-i\phi}\sin(\theta/2)\langle z_{-}|)(\cos(\theta/2)|z_{+}\rangle-e^{i\phi}\sin(\theta/2)|z_{-}\rangle) \\
& = \cos^{2}(\theta/2)-\sin^{2}(\theta/2) \\
& = \cos\theta.
\end{align}

Com isso obteremos:
\begin{align}
\langle\sigma_{x}(t)\rangle & = \langle\psi_{0}^{S}|\sigma_{x}^{H}|\psi_{0}^{S}\rangle \\
& = \langle\psi_{0}^{S}|(\sigma_{x}^{S}\cos\omega t + \sigma_{y}^{S}\sin\omega t)|\psi_{0}^{S}\rangle \\ 
& = \langle\sigma_{x}^{S}\rangle_{n_{+}}\cos\omega t + \langle\sigma_{y}^{S}\rangle_{n_{+}}\sin\omega t \\
& = \sin\theta\cos\phi\cos\omega t + \sin\theta\sin\phi\sin\omega t \\
& = \sin\theta\cos(\phi+\omega t), \\
\langle\sigma_{y}(t)\rangle & = \langle\psi_{0}^{S}|\sigma_{y}^{H}|\psi_{0}^{S}\rangle \\
& = \langle\psi_{0}^{S}|(-\sigma_{x}^{S}\sin\omega t + \sigma_{y}^{S}\cos\omega t)|\psi_{0}^{S}\rangle \\ 
& = -\langle\sigma_{x}^{S}\rangle_{n_{+}}\sin\omega t + \langle\sigma_{y}^{S}\rangle_{n_{+}}\cos\omega t \\
& = -\sin\theta\cos\phi\sin\omega t + \sin\theta\sin\phi\cos\omega t \\
& = \sin\theta\sin(\phi+\omega t), \\
\langle\sigma_{z}(t)\rangle & = \langle\psi_{0}^{S}|\sigma_{z}^{H}|\psi_{0}^{S}\rangle \\
& = \langle\psi_{0}^{S}|\sigma_{z}^{S}|\psi_{0}^{S}\rangle \\ 
& = \cos\theta,
\end{align}
que são os mesmos resultados que obtivemos na RS.

OBS. Vale mencionar que, na RH, se usamos as equações de Heisenberg, não obtemos as probabilidades, mas somente os valores médios.

### Precessão de spin para $[H_t,H_{t'}]=0$

Vamos considerar um quanton de spin 1/2 sujeito a um campo magnético constante no espaço mas dependente do tempo
$$\vec{B}_t = B_t \hat{k}.$$
Lembrando, o momento de dipolo magnético pode ser escrito como 
$$\vec{\mu} = \frac{g q}{2m}\vec{S} = \frac{\hbar g q}{4m}\vec{\sigma}.$$
Com isso, o hamiltoniano de interação do momento de dipolo magnético com o campo magnético é
\begin{align}
H_t & = -\vec{\mu}\cdot\vec{B}_t = -\frac{\hbar g q}{4m}\vec{\sigma}\cdot B_t \hat{k} \\
& =: -\frac{\hbar\omega_t}{2}\sigma_z, 
\end{align}
em que definimos a frequência
$$\omega_t := \frac{g qB_t}{2m}.$$
Note que essa frequência depende da forma de $B_t$. Se este e.g. varia senoidalmente, a frequência também varia dessa forma. Se e.g. $B_t = B_0 t$, então a frequência aumentará linearmente com o tempo. Mas para as contas que faremos na sequência, não precisaremos especificar a forma da dependência do campo magnético com o tempo.

Começamos verificando que esse hamiltoniano comuta em tempos diferentes:
\begin{align}
[H_t,H_{t'}] & = \big[-\frac{\hbar\omega_t}{2}\sigma_z,-\frac{\hbar\omega_{t'}}{2}\sigma_z\big] = \frac{\hbar^2 \omega_t\omega_{t'}}{4}[\sigma_z,\sigma_z] \\
& = 0.
\end{align}

Vimos que nesta situação o operador de evolução temporal 
$$U_t = e^{-\frac{i}{\hbar}\int_{0}^{t}H_{t'}dt'}$$
é solução da equação de Schrödiner
$i\hbar\partial_t U_t = H_t U_t.$
Com isso, vem que
\begin{align}
U_t & = e^{-\frac{i}{\hbar}\int_{0}^{t}(-\hbar\omega_{t'}\sigma_z /2)dt'} = e^{i\big(\int_{0}^{t}\omega_{t'}dt'\big)\sigma_z /2} \\
& =: e^{if_t \sigma_z /2} \\
& = e^{if_t /2}|z_+\rangle\langle z_+| + e^{-if_t /2}|z_-\rangle\langle z_-|,
\end{align}
em que definimos a função escalar do tempo
$$f_t := \int_{0}^{t}\omega_{t'}dt'.$$

Com isso, começando com um estado de spin geral na direção $\hat{n}$, i.e., 
$$|\psi_0\rangle = |n_+\rangle = \cos(\theta/2)|z_+\rangle + e^{i\phi}\sin(\theta/2)|z_-\rangle,$$
teremos que o estado evoluído será dado por
\begin{align}
|\psi_t\rangle & = U_t|\psi_0\rangle \\
& = \big(e^{if_t /2}|z_+\rangle\langle z_+| + e^{-if_t /2}|z_-\rangle\langle z_-|\big)\big(\cos(\theta/2)|z_+\rangle + e^{i\phi}\sin(\theta/2)|z_-\rangle\big) \\
& = e^{if_t /2}\cos(\theta/2)|z_+\rangle + e^{-if_t /2}e^{i\phi}\sin(\theta/2)|z_-\rangle \\
& = e^{if_t /2}\big(\cos(\theta/2)|z_+\rangle + e^{i(\phi-f_t)}\sin(\theta/2)|z_-\rangle\big) \\
& \equiv \cos(\theta/2)|z_+\rangle + e^{i(\phi-f_t)}\sin(\theta/2)|z_-\rangle,
\end{align}
em que desconsideramos a fase global $e^{if_t /2}$. Se usarmos a representação de estados quânticos de sistemas de 2 níveis na esfera de Bloch (isomorfismo $SU(2)\leftrightarrow SO(3)$), vemos que a representação deste estado em $\mathbb{R}^3$ precessa em torno da direção $z$ positiva, com o ângulo azimutal dado por
$$\phi_t = \phi_0 - f_t.$$

Por exemplo, se o campo magnético aumenta linearmente com o tempo,
$$B_t = B_0 t,$$ 
então 
$$\omega_t = \frac{gqB_0}{2m}t=\omega_0 t$$
e
$$f_{t} = \int_{0}^{t}\omega_0 t'dt' = \frac{\omega_0 t^2}{2}.$$
Assim
$$\phi_t = \phi_0 - f_t = \phi_0 - \frac{\omega_0 t^2}{2}.$$
