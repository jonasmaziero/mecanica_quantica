# Oscilador harmônico quântico 1D

Hamiltoniano
\begin{equation}
H = \frac{P^{2}}{2m} + \frac{m\omega^{2}X^{2}}{2}.
\end{equation}

## Diagonalização via operadores escada
Definimos os operadores escada
\begin{equation}
a := \sqrt{\frac{m\omega}{2\hbar}}\left(X + \frac{i}{m\omega}P\right) \Rightarrow a^{\dagger} := \sqrt{\frac{m\omega}{2\hbar}}\left(X - \frac{i}{m\omega}P\right),
\end{equation}
que nos fornecem
\begin{equation}
X = \sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}+a) \text{ e } P = i\sqrt{\frac{\hbar m\omega}{2}}(a^{\dagger}-a).
\end{equation}

Aqui usaremos a relação canônica de comutação entre posição e momento
\begin{equation}
[X,P]=i\hbar\mathbb{I},
\end{equation}
que será demonstrada posteriormente, para verificar que
\begin{align}
[a,a^{\dagger}] & = \frac{m\omega}{2\hbar}\left[X + \frac{i}{m\omega}P,X - \frac{i}{m\omega}P\right] \\
& = \frac{m\omega}{2\hbar}\left(\mathbb{0} - \frac{i}{m\omega}[X,P] + \frac{i}{m\omega}[P,X] + \mathbb{0}\right) \\
& = \frac{m\omega}{2\hbar}\frac{-2i}{m\omega}[X,P] = \frac{-i}{\hbar}i\hbar\mathbb{I} \\
& = \mathbb{I}=aa^{\dagger}-a^{\dagger}a.
\end{align}

Assim, o Hamiltoniano fica escrito como
\begin{align}
H & = -\frac{\hbar m\omega}{2}(a^{\dagger}-a)(a^{\dagger}-a)/2m + \frac{m\omega^{2}}{2}\frac{\hbar}{2m\omega}(a^{\dagger}+a)(a^{\dagger}+a)\\
& = \frac{\hbar\omega}{4}\left(-(a^{\dagger})^{2}+a^{\dagger}a+aa^{\dagger}-a^{2} + (a^{\dagger})^{2}+a^{\dagger}a+aa^{\dagger}+a^{2}\right) \\
& = \frac{\hbar\omega}{2}(a^{\dagger}a+aa^{\dagger}) = \frac{\hbar\omega}{2}(a^{\dagger}a+\mathbb{I}+a^{\dagger}a) \\
& = \hbar\omega(N+\mathbb{I}/2), 
\end{align}
onde definimos o _operador número_ $N:=a^{\dagger}a$.

## Diagonalização do Hamiltoniano
Vamos escrever
\begin{equation}
N|n\rangle=n|n\rangle
\end{equation}
e obter os possíveis valores de $n$. Primeiramente usamos a positividade da norma de um vetor qualquer do espaço de Hilbert associado
\begin{align}
&||a|n\rangle||\ge 0 \Rightarrow 0\le\langle n|a^{\dagger}a|n\rangle =: \langle n|N|n\rangle = n\langle n|n\rangle=n \\
& \Rightarrow n \ge 0.
\end{align}
A principal observação a ser feita na sequência é sobre a ação do operadores escada nos autovetores de $N$, que também são autovetores de $H$. Começamos com
\begin{align}
N(a|n\rangle) & = a^{\dagger}aa|n\rangle = (aa^{\dagger}-\mathbb{I})a|n\rangle \\
& = aa^{\dagger}a|n\rangle-a|n\rangle = aN|n\rangle-a|n\rangle = an|n\rangle-a|n\rangle \\
& = (n-1)(a|n\rangle).
\end{align}
Ou seja, $a|n\rangle$ também é autovetor de $N$, mas correspondente ao autovalor $n-1$. Note que o estado $|n-1\rangle$ é autovetor de $N$ correspondente a este mesmo autovalor. Portanto devemos ter
\begin{equation}
a|n\rangle\propto|n-1\rangle =: c|n-1\rangle.
\end{equation}
Para que possam representar estados físicamente aceitáveis, esses vetores devem ser normalizados. Vem então que
\begin{align}
& \langle n| a^{\dagger}a|n\rangle = \langle n-1|c^{*}c|n-1\rangle, \\
& \langle n|N|n\rangle = |c|^{2}\langle n-1|n-1\rangle, \\
& n\langle n|n\rangle = n = |c|^{2} \Rightarrow |c|=\sqrt{n},
\end{align}
que nos fornece a ação do operador de abaixamento na base número (assumimos $c\in\mathbb{R}$):
\begin{equation}
a|n\rangle = \sqrt{n}|n-1\rangle.
\end{equation}

__Exercício:__ Verifique que $N(a^{\dagger}|n\rangle)=(n+1)(a^{\dagger}|n\rangle)$ e $a^{\dagger}|n\rangle=\sqrt{n+1}|n+1\rangle$.

Já sabemos que $n\ge 0$ e que $a$ leva $|n\rangle$ em um autovetor de $N$ "diminuído por uma unidade". Podemos usar esses fatos para argumentar que $n$ deve ser inteiro. O menor valor possível de $n$ é $0$. Então se $a$ atua no autovetor $|n=0\rangle\equiv|0\rangle$ devemos obter um vetor que não é um estado possível, que é o vetor nulo, i.e., \begin{equation}
a|0\rangle=|\oslash\rangle.
\end{equation}
Agora, se $n$ pudesse ser real, teríamos, por exemplo, $a|0.25\rangle=|-0.75\rangle$, que não é um valor possível de $n$. Concluímos então que $n\in\mathbb{Z}_{\ge 0}$. Por outro lado, $a^{\dagger}$ pode aumentar o autovalor $n$ indefinidamente. Resumindo
\begin{align}
& N|n\rangle=n|n\rangle \text{ com } n=0,1,2,3,\cdots, \\
& H|n\rangle = E_{n}|n\rangle=\hbar\omega(n+1/2)|n\rangle.
\end{align}

### Valores médios e desvios padrões
Consideremos o oscilador preparado em um estado com energia definida. Para a posição teremos
\begin{align}
\langle X\rangle_{n} &= \langle n|X|n\rangle = \langle n|\sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}+a)|n\rangle \\
& = \sqrt{\frac{\hbar}{2m\omega}}(\langle n|a^{\dagger}|n\rangle+\langle n|a|n\rangle) \\
& = \sqrt{\frac{\hbar}{2m\omega}}(\langle n|\sqrt{n+1}|n+1\rangle+\langle n|\sqrt{n}|n-1\rangle) \\
& = 0.
\end{align}
e
\begin{align}
\langle X^{2}\rangle_{n} &= \frac{\hbar}{2m\omega}\langle n|(a^{\dagger}+a)(a^{\dagger}+a)|n\rangle \\
& = \frac{\hbar}{2m\omega}(\langle n|(a^{\dagger})^{2}|n\rangle+\langle n|a^{2}|n\rangle + \langle n|a^{\dagger}a|n\rangle + \langle n|aa^{\dagger}|n\rangle) \\
& = \frac{\hbar}{2m\omega}(\langle n|\sqrt{n+1}a^{\dagger}|n+1\rangle+\langle n|\sqrt{n}a|n-1\rangle + \langle n|N|n\rangle + \langle n|(a^{\dagger}a+\mathbb{I})|n\rangle) \\
& = \frac{\hbar}{2m\omega}(\langle n|\sqrt{n+1}\sqrt{n+2}|n+2\rangle+\langle n|\sqrt{n}\sqrt{n-1}|n-2\rangle + \langle n|N|n\rangle + \langle n|(N+\mathbb{I})|n\rangle) \\
& = \frac{\hbar}{2m\omega}(0+0 + n + (n+1)) \\
& = \frac{\hbar}{2m\omega}(2n+1).
\end{align}

Var(X)_{0}Var(P)_{0}=\frac{\hbar^{2}}{4}.
\end{equation}

### Dinâmica
O operador de evolução temporal é $U_{t}=e^{-iHt/\hbar}=\sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\langle n|$ com $E_{n}=\frac{\hbar\omega}{2}(n+1/2)$.
#### Dinâmica na representação de Schrödinger
Se o estado preparado for um autovetor de $H$, este só ganha uma fase, i.e., $|n\rangle\rightarrow e^{-iE_{n}t/\hbar}|n\rangle$. Consideremos então que o estado inicial é uma combinação linear qualquer $|\psi_{0}\rangle = \sum_{m}c_{m}|m\rangle$. Teremos assim que
\begin{align}
|\psi_{t}\rangle & = U_{t}|\psi_{0}\rangle = \sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\langle n|\sum_{m}c_{m}|m\rangle \\
& = \sum_{m}c_{m}\sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\langle n|m\rangle = \sum_{m}c_{m}\sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\delta_{n,m} \\
& = \sum_{m}c_{m}e^{-iE_{m}t/\hbar}|m\rangle.
\end{align}
Ou seja, o estado $|m\rangle$ ganha uma fase global $e^{-iE_{m}t/\hbar}$ proporcinal à energia daquele estado.

#### Dinâmica na representação de Heisenberg
Para esse sistema, nessa representação, é conveniente olhar para a evolução temporal dos operadores escada. Aqui, como na representação de Schrödinger $H_{S}=\hbar\omega(a_{S}^{\dagger}a_{S}+\mathbb{I}/2)$, na representação de Heisenberg $H_{H}=\hbar\omega(a_{H}^{\dagger}a_{H}+\mathbb{I}/2)$. Vamos para a equação de Heisenberg:
\begin{align}
i\hbar\partial_{t}a_{H} &= [a_{H},H_{H}] = \hbar\omega([a_{H},a_{H}^{\dagger}a_{H}]+[a_{H},\mathbb{I}]/2) \\
& = \hbar\omega(a_{H}^{\dagger}[a_{H},a_{H}] + [a_{H},a_{H}^{\dagger}]a_{H} + \mathbb{O}) = \hbar\omega(\mathbb{O} + \mathbb{I}a_{H} + \mathbb{O}) \\
& = \hbar\omega a_{H}.
\end{align}
A solução é
\begin{equation}
a_{H} = e^{-i\omega t}a_{S}.
\end{equation}
