# Oscilador harmônico quântico 1D
O potencial harmônico clássico unidimensional, $U=kx^{2}/2$, correspondente à força restauradora $F_{x}=-\partial_{x}U=-kx$ (e usando $\omega=\sqrt{k/m}$), nos fornece, depois da quantização $x\rightarrow X$ e $p\rightarrow P$, o Hamiltoniano quântico para o oscilador harmônico quântico (OHQ) 1D:
\begin{equation}
H_{ohq} = \frac{P^{2}}{2m} + \frac{m\omega^{2}X^{2}}{2}.
\end{equation}

## Operadores escada
Definimos os operadores escada como uma combinação linear dos operadores posição $X$ e momento linear $P$:
\begin{equation}
a := \sqrt{\frac{m\omega}{2\hbar}}\left(X + \frac{i}{m\omega}P\right) \Rightarrow a^{\dagger} := \sqrt{\frac{m\omega}{2\hbar}}\left(X - \frac{i}{m\omega}P\right).
\end{equation}

Podemos manipular essas equações para obter:
\begin{equation}
X = \sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}+a) \text{ e } P = i\sqrt{\frac{\hbar m\omega}{2}}(a^{\dagger}-a).
\end{equation}

Aqui usaremos a relação canônica de comutação entre posição e momento linear (que obteremos mais adiante),
\begin{equation}
[X,P]=i\hbar\mathbb{I},
\end{equation}
para verificar que
\begin{align}
[a,a^{\dagger}] & = \frac{m\omega}{2\hbar}\left[X + \frac{i}{m\omega}P,X - \frac{i}{m\omega}P\right] \\
& = \frac{m\omega}{2\hbar}\left(\mathbb{0} - \frac{i}{m\omega}[X,P] + \frac{i}{m\omega}[P,X] + \mathbb{0}\right) \\
& = \frac{m\omega}{2\hbar}\frac{-2i}{m\omega}[X,P] = \frac{-i}{\hbar}i\hbar\mathbb{I} \\
& = \mathbb{I} \\ 
& =aa^{\dagger}-a^{\dagger}a.
\end{align}

Assim, o Hamiltoniano fica escrito como
\begin{align}
H_{ohq} & = -\frac{\hbar m\omega}{2}(a^{\dagger}-a)(a^{\dagger}-a)/2m + \frac{m\omega^{2}}{2}\frac{\hbar}{2m\omega}(a^{\dagger}+a)(a^{\dagger}+a)\\
& = \frac{\hbar\omega}{4}\left(-(a^{\dagger})^{2}+a^{\dagger}a+aa^{\dagger}-a^{2} + (a^{\dagger})^{2}+a^{\dagger}a+aa^{\dagger}+a^{2}\right) \\
& = \frac{\hbar\omega}{2}(a^{\dagger}a+aa^{\dagger}) = \frac{\hbar\omega}{2}(a^{\dagger}a+\mathbb{I}+a^{\dagger}a) \\
& = \hbar\omega(N+\mathbb{I}/2), 
\end{align}
onde definimos o _operador número_, que é Hermintiano, como:
\begin{equation}
N:=a^{\dagger}a.
\end{equation}

## Diagonalização do Hamiltoniano
Vamos definir
\begin{equation}
N|n\rangle=n|n\rangle
\end{equation}
e obter os possíveis valores de $n$. Primeiramente usamos a positividade da norma de um vetor qualquer do espaço de Hilbert associado:
\begin{align}
&||a|n\rangle||^{2}\ge 0 \Rightarrow 0\le(a|n\rangle,a|n\rangle)=(|n\rangle,a^{\dagger}a|n\rangle) =(|n\rangle,N|n\rangle) = n\langle n|n\rangle=n \\
& \Rightarrow n \ge 0.
\end{align}

A principal observação a ser feita na sequência é sobre a ação dos operadores escada nos autovetores de $N$, que também são autovetores de $H$. Começamos com
\begin{align}
N(a|n\rangle) & = a^{\dagger}aa|n\rangle = (aa^{\dagger}-\mathbb{I})a|n\rangle \\
& = aa^{\dagger}a|n\rangle-a|n\rangle = aN|n\rangle-a|n\rangle = an|n\rangle-a|n\rangle \\
& = (n-1)(a|n\rangle).
\end{align}
Ou seja, $a|n\rangle$ também é autovetor de $N$, mas correspondente ao autovalor $n-1$. Note que o estado $|n-1\rangle$ é autovetor de $N$ correspondente a este mesmo autovalor, e portanto devemos ter
\begin{equation}
a|n\rangle\propto|n-1\rangle =: c|n-1\rangle.
\end{equation}

Para que possam representar estados físicamente aceitáveis, esses vetores devem ser normalizados. Vem então que
\begin{align}
& \langle n| a^{\dagger}a|n\rangle = \langle n-1|c^{*}c|n-1\rangle, \\
& \langle n|N|n\rangle = |c|^{2}\langle n-1|n-1\rangle, \\
& n\langle n|n\rangle = n = |c|^{2} \Rightarrow |c|=\sqrt{n},
\end{align}
que nos fornece a ação do operador de abaixamento na base número (assumimos $c\in\mathbb{R}$):
\begin{equation}
a|n\rangle = \sqrt{n}|n-1\rangle.
\end{equation}

Já sabemos que $n\ge 0$ e que $a$ leva $|n\rangle$ em um autovetor de $N$ "diminuído por uma unidade". Podemos usar esses fatos para argumentar que $n$ deve ser inteiro. O menor valor possível de $n$ é $0$. Então se $a$ atua no autovetor $|n=0\rangle\equiv|0\rangle$, devemos obter um vetor que não é um estado possível, que é o vetor nulo, i.e., \begin{equation}
a|0\rangle=|\oslash\rangle.
\end{equation}
Agora, se $n$ pudesse ser real, teríamos, por exemplo, $a|0.5\rangle=|-0.5\rangle$, que não é um valor possível de $n$. Concluímos então que $n\in\mathbb{Z}_{\ge 0}$. 

Também temos que
\begin{equation}
a^{\dagger}|n\rangle = \sqrt{n+1}|n+1\rangle.
\end{equation}
Em contraste com $a$, $a^{\dagger}$ pode aumentar o autovalor $n$ indefinidamente. Não há argumento matemático impondo um limite superior para $n$. Resumindo
\begin{align}
& N|n\rangle=n|n\rangle \text{ com } n=0,1,2,3,\cdots, \\
& H_{ohq}|n\rangle = E_{n}|n\rangle=\hbar\omega(n+1/2)|n\rangle.
\end{align}
Vemos, no entanto, que há uma restrição física para o limite superior de $n$, e que é imposta pela energia máxima que o OHQ poderá ter. Além disso, a restrição $n\ge 0$ implica em uma energia mínima não nula dos OHQ, a chamada energia de ponto zero, que apresenta o fato que todo OHQ está sempre oscilando.

Abaixo está mostrado o diagrama de níveis de energia para o OHQ1D.

```python
xmin = 0; xmax = 1
for j in range(0,5): # desenha as linhas
    plt.hlines(j,xmin,xmax,color='blue'); 
for j in range(0,4): # valores de n e de En, e hw e <->
    plt.text(-0.13,j,j,color='black',fontsize=13)
    plt.text(1.01,j, str(2*j+1)+r'$\hbar\omega/2$',color='black',fontsize=13)
for j in range(0,3):
    plt.annotate("", xy=(0.5,j), xytext=(0.5,j+1), arrowprops=dict(arrowstyle="<->"))
    plt.text(0.51,(2*j+1)/2,r'$\hbar\omega$',fontsize=13)
plt.text(0.45,3.5,r'$\vdots$',fontsize=20)
plt.text(-0.13,4,r'$n$',color='black',fontsize=13); plt.text(1.01,4,r'$E_{n}=\hbar\omega(n+1/2)$',fontsize=13)
plt.axis('off'); plt.xlim(0,2); plt.ylim(-0.05,4.05); plt.show()
```

## Valores médios e desvios padrões de $X$ e $P$ para $|\psi\rangle=|n\rangle$
Consideremos o OHQ preparado em um estado com energia definida. Para a posição teremos
\begin{align}
\langle X\rangle_{n} &= \langle n|X|n\rangle = \langle n|\sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}+a)|n\rangle = \sqrt{\frac{\hbar}{2m\omega}}(\langle n|a^{\dagger}|n\rangle+\langle n|a|n\rangle) \\
& = \sqrt{\frac{\hbar}{2m\omega}}(\langle n|\sqrt{n+1}|n+1\rangle+\langle n|\sqrt{n}|n-1\rangle) \\
& = 0.
\end{align}
e
\begin{align}
\langle X^{2}\rangle_{n} &= \frac{\hbar}{2m\omega}\langle n|(a^{\dagger}+a)(a^{\dagger}+a)|n\rangle = \frac{\hbar}{2m\omega}(\langle n|(a^{\dagger})^{2}|n\rangle+\langle n|a^{2}|n\rangle + \langle n|a^{\dagger}a|n\rangle + \langle n|aa^{\dagger}|n\rangle) \\
& = \frac{\hbar}{2m\omega}(\langle n|\sqrt{n+1}a^{\dagger}|n+1\rangle+\langle n|\sqrt{n}a|n-1\rangle + \langle n|N|n\rangle + \langle n|(a^{\dagger}a+\mathbb{I})|n\rangle) \\
& = \frac{\hbar}{2m\omega}(\langle n|\sqrt{n+1}\sqrt{n+2}|n+2\rangle+\langle n|\sqrt{n}\sqrt{n-1}|n-2\rangle + \langle n|N|n\rangle + \langle n|(N+\mathbb{I})|n\rangle) \\
& = \frac{\hbar}{2m\omega}(0+0 + n + (n+1)) \\
& = \frac{\hbar}{2m\omega}(2n+1).
\end{align}

Analogamente, para o momento linear teremos:
\begin{equation}
\langle P\rangle_{n} = 0 \text{ e } \langle P^{2}\rangle_{n}=\frac{\hbar m\omega}{2}(2n+1).
\end{equation}
Portanto, a incerteza sobre a posição e sobre o momento linear aumentam com $n$. Juntando os resultados acima vemos que
\begin{align}
& Var(X)_{n}Var(P)_{n}=\frac{\hbar}{2m\omega}(2n+1)\frac{\hbar m\omega}{2}(2n+1) = \frac{\hbar^{2}}{4}(2n+1)^{2}, \\
& \therefore \sigma_{X}^{n}\sigma_{P}^{n} = \hbar(2n+1)/2.
\end{align}
Temos assim que o estado fundamental do oscilador harmônico, $n=0$, atinge o limite inferior para o produto das incertezas da posição e momento linear (discutiremos relações de incerteza mais adiante):
\begin{equation}
Var(X)_{0}Var(P)_{0}=\frac{\hbar^{2}}{4}.
\end{equation}
i.e., $|0\rangle$ é um estado de mínima incerteza.


## Valores médios e desvios padrões de $X$ e $P$ para $|\psi\rangle=\sum_{n}c_{n}|n\rangle$
Para a posição
\begin{align}
\langle X\rangle_{\psi} & = \sqrt{\frac{\hbar}{2m\omega}}\langle\psi|(a^{\dagger}+a)\sum_{n}c_{n}|n\rangle = \sqrt{\frac{\hbar}{2m\omega}}\langle\psi|\sum_{n}c_{n}(a^{\dagger}|n\rangle+a|n\rangle) \\
& = \sqrt{\frac{\hbar}{2m\omega}}\sum_{m}c_{m}^{*}\langle m|\sum_{n}c_{n}(\sqrt{n+1}|n+1\rangle+\sqrt{n}|n-1\rangle) \\
& = \sqrt{\frac{\hbar}{2m\omega}}\sum_{m,n}c_{m}^{*}c_{n}(\sqrt{n+1}\langle m|n+1\rangle+\sqrt{n}\langle m|n-1\rangle) \\
& = \sqrt{\frac{\hbar}{2m\omega}}\sum_{m,n}c_{m}^{*}c_{n}(\sqrt{n+1}\delta_{m,n+1}+\sqrt{n}\delta_{m,n-1}) \\
& = \sqrt{\frac{\hbar}{2m\omega}}\sum_{n}c_{n}\left(c_{n+1}^{*}\sqrt{n+1} + c_{n-1}^{*}\sqrt{n}\right).
\end{align}

Por exemplo, para $|\psi\rangle=(|0\rangle+|1\rangle)/\sqrt{2}$ teremos $c_{0}=c_{1}=1/\sqrt{2}, c_{n>1}=0$ e assim 
\begin{align}
\langle X\rangle_{\psi} & = \sqrt{\frac{\hbar}{2m\omega}}\left(c_{0}\left(c_{0+1}^{*}\sqrt{0+1} + c_{0-1}^{*}\sqrt{0}\right) + c_{1}\left(c_{1+1}^{*}\sqrt{1+1} + c_{1-1}^{*}\sqrt{1}\right) + 0 \right) \\
& = \sqrt{\frac{\hbar}{2m\omega}}.
\end{align}

Do mesmo modo que para $\langle X\rangle_{\psi}$, pode-se mostrar que 
\begin{equation}
\langle P\rangle_{\psi} = \cdots = i\sqrt{\frac{\hbar m\omega}{2}}\sum_{n}c_{n}\left(c_{n+1}^{*}\sqrt{n+1} - c_{n-1}^{*}\sqrt{n}\right).
\end{equation}

OBS. Vemos assim que cada par $x_{0},p_{0}$ usado para fazer o gráfico apresentado na próxima seção corresponde a uma (ou mais) combinação linear dos estados da base número.

## Dinâmica gerada pelo Hamiltoniano do OHQ
Como $H_{ohq}=\sum_{n=0}^{\infty}E_{n}|n\rangle\langle n|$, com $E_{n}=\frac{\hbar\omega}{2}(n+1/2)$, o operador de evolução temporal é
\begin{equation}
U_{t}=e^{-iH_{ohq}t/\hbar}=\sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\langle n|.
\end{equation}

### Dinâmica do OHQ na representação de Schrödinger
Se o estado preparado for um autovetor de $H_{ohq}$, este só ganha uma fase, i.e., $|n\rangle\rightarrow e^{-iE_{n}t/\hbar}|n\rangle$. Consideremos então que o estado inicial é uma combinação linear qualquer $|\psi_{0}\rangle = \sum_{m}c_{m}|m\rangle$. Teremos assim que
\begin{align}
|\psi_{t}\rangle & = U_{t}|\psi_{0}\rangle = \sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\langle n|\sum_{m}c_{m}|m\rangle \\
& = \sum_{m}c_{m}\sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\langle n|m\rangle = \sum_{m}c_{m}\sum_{n=0}^{\infty}e^{-iE_{n}t/\hbar}|n\rangle\delta_{n,m} \\
& = \sum_{m}c_{m}e^{-iE_{m}t/\hbar}|m\rangle.
\end{align}
Ou seja, o estado $|m\rangle$ ganha uma fase $e^{-iE_{m}t/\hbar}$ proporcional à energia daquele estado.

### Dinâmica do OHQ na representação de Heisenberg
Para esse sistema, nessa representação, é conveniente olhar para a evolução temporal dos operadores escada. Como na representação de Schrödinger $H_{ohq}^{S}=\hbar\omega(a_{S}^{\dagger}a_{S}+\mathbb{I}/2)$, com $a_{S}\equiv a$, na representação de Heisenberg $H_{ohq}^{H}=\hbar\omega(a_{H}^{\dagger}a_{H}+\mathbb{I}/2)$. Vamos para a equação de Heisenberg:
\begin{align}
i\hbar\partial_{t}a_{H} &= [a_{H},H_{ohq}^{H}] + \mathbb{O} = \hbar\omega([a_{H},a_{H}^{\dagger}a_{H}]+[a_{H},\mathbb{I}]/2) \\
& = \hbar\omega(a_{H}^{\dagger}[a_{H},a_{H}] + [a_{H},a_{H}^{\dagger}]a_{H} + \mathbb{O}) = \hbar\omega(\mathbb{O} + \mathbb{I}a_{H} + \mathbb{O}) \\
& = \hbar\omega a_{H}, \\
& \therefore \partial_{t}a_{H} = -i\omega a_{H}.
\end{align}
A solução é
\begin{equation}
a_{H} = e^{-i\omega t}a,
\end{equation}
onde usamos $a_{H}(t=0)\equiv a_{S}\equiv a$. Analogamente, $a_{H}^{\dagger} = e^{i\omega t}a_{S}^{\dagger}$. Com isso vem que
\begin{align}
X_{t}^{H} & = \sqrt{\frac{\hbar}{2m\omega}}(e^{i\omega t}a^{\dagger}+e^{-i\omega t}a) = \cdots = X\cos\omega t + \frac{P}{m\omega}\sin\omega t, \\
P_{t}^{H} & = i\sqrt{\frac{\hbar m\omega}{2}}(e^{i\omega t}a^{\dagger}-e^{-i\omega t}a) = \cdots = -m\omega X\sin\omega t + P\cos\omega t,
\end{align}
As equações acima podem ser escritas como uma única equação matricial:
\begin{equation}
\begin{bmatrix} X_{t}^{H} \\ P_{t}^{H}/m\omega \end{bmatrix} = \begin{bmatrix} \cos\omega t & \sin\omega t \\ -\sin\omega t & \cos\omega t \end{bmatrix}\begin{bmatrix} X \\ P/m\omega \end{bmatrix}
\end{equation}
Vemos assim que a dinâmica sob o Hamiltoniano do OHQ gira os valores médios da posição e do momento no 'espaço de fase' com coordenadas $(\langle X_{t}^{H}\rangle_{\psi_{0}},\langle P_{t}^{H}\rangle_{\psi_{0}}/m\omega)$. O raio do círculo é igual a norma do vetor $\begin{bmatrix} X_{t}^{H} & P_{t}^{H}/m\omega \end{bmatrix}^{T}$, que é $\sqrt{\langle X\rangle^{2}+\langle P/m\omega\rangle^{2}}$. Veja a figura abaixo.

```python
def ohq_dyn(x0,p0,tmax):
    fig = plt.figure(figsize=(5,5))
    w = 1; t = np.arange(0,tmax,0.01); x = np.cos(w*t)*x0 + np.sin(w*t)*p0; y = -np.sin(w*t)*x0 + np.cos(w*t)*p0
    plt.plot(x,y,linestyle='dashed'); plt.xlim(-1,1); plt.ylim(-1,1)
    plt.xlabel(r'$\langle X_{t}^{H}\rangle$',fontsize=13)
    plt.ylabel(r'$\langle P_{t}^{H}\rangle/m\omega$',fontsize=13)
    plt.annotate(r'$t=$'+str(tmax)[:5], xy=(0.6,0.8), xytext=(0.6,0.8), fontsize=12)
    xm = cos(w*tmax)*x0 + sin(w*tmax)*p0; ym = -sin(w*tmax)*x0 + cos(w*tmax)*p0
    plt.annotate(r'$\bigcirc$', xy=(xm,ym), xytext=(xm,ym), fontsize=12); plt.show()
interactive(ohq_dyn, x0 = (0,1,0.1), p0 = (0,1,0.1), tmax = (0,20,0.05))
```

# Estados coerentes do OHQ
Os estados coerentes foram importantes para o desenvolvimento de uma área muito importante da Física, a Óptica Quântica (veja e.g. [https://www.nobelprize.org/prizes/physics/2005/glauber/lecture/]) e são as seguintes combinações lineares dos estados da base número $\{|n\rangle\}$:
\begin{equation}
|\alpha\rangle = e^{-|\alpha|^{2}/2}\sum_{n=0}^{\infty}\frac{\alpha^{n}}{\sqrt{n!}}|n\rangle,
\end{equation}
onde $\alpha\in\mathbb{C}$. Esses estados são autovetores do operador de abaixamento (que não é Hermitiano):
\begin{align}
a|\alpha\rangle & = e^{-|\alpha|^{2}/2}\sum_{n=0}^{\infty}\frac{\alpha^{n}}{\sqrt{n!}}a|n\rangle  \\
& = e^{-|\alpha|^{2}/2}\left(0+\sum_{n=1}^{\infty}\frac{\alpha\alpha^{n-1}}{\sqrt{n((n-1)!)}}\sqrt{n}|n-1\rangle\right) \\ 
& = \alpha e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}|m\rangle \\ 
& = \alpha|\alpha\rangle.
\end{align}

### Distribuição de probabilidades para a energia
O valor médio do operador número nesse estado é:
\begin{align}
\langle N\rangle_{\alpha} & = \langle\alpha|N|\alpha\rangle=\langle\alpha|a^{\dagger}a|\alpha\rangle = \langle\alpha|\alpha^{*}\alpha|\alpha\rangle = |\alpha|^{2}\langle\alpha|\alpha\rangle \\ 
& = |\alpha|^{2}.
\end{align}

A probabilidade do OHQ preparado em um estado coerente ter energia $E_{n}$ é:
\begin{align}
Pr(E_{n}|\alpha) & = |\langle n|\alpha\rangle|^{2} = \left|\langle n|e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}|m\rangle\right|^{2} \\
& = \left|e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}\langle n|m\rangle\right|^{2} = \left|e^{-|\alpha|^{2}/2}\sum_{m=0}^{\infty}\frac{\alpha^{m}}{\sqrt{m!}}\delta_{n,m}\right|^{2} \\
& = \left|\frac{e^{-|\alpha|^{2}/2}\alpha^{n}}{\sqrt{n!}}\right|^{2} = \frac{e^{-|\alpha|^{2}}|\alpha|^{2n}}{n!} \\ 
& = \frac{e^{-\langle N\rangle_{\alpha}}\langle N\rangle_{\alpha}^{n}}{n!},
\end{align}
que é denominada __distribuição Poissoniana__ de probabilidades.

```python
def poisson(Nm):
    N = 11; x = np.zeros(N); pr = np.zeros(N)
    for n in range(0,N):
        x[n] = n; pr[n] = (math.exp(-Nm)*Nm**n)/np.math.factorial(n)
    plt.plot(x,pr); plt.xlim(0,N-1); plt.ylim(0,1)
    plt.xlabel('n',fontsize=13); plt.ylabel(r'$Pr(E_{n}|\alpha)$',fontsize=13)
    plt.annotate(r'$\langle N\rangle=$'+str(Nm), xy=(8,0.9), xytext=(8,0.9)); plt.show()
interactive(poisson, Nm = (0,5,0.025))
```

### Estados coerentes são os "mais clássicos" dos estados quânticos
Teremos 
\begin{align}
\langle X\rangle_{\alpha} & = \langle\alpha|\sqrt{\hbar/2m\omega}(a^{\dagger}+a)|\alpha\rangle = \sqrt{\hbar/2m\omega}\left(\langle\alpha|a^{\dagger}|\alpha\rangle+\langle\alpha|a|\alpha\rangle\right) \\
& = \sqrt{\hbar/2m\omega}\left((a|\alpha\rangle,|\alpha\rangle)+(|\alpha\rangle,a|\alpha\rangle)\right) = \sqrt{\hbar/2m\omega}\left((\alpha|\alpha\rangle,|\alpha\rangle)+(|\alpha\rangle,\alpha|\alpha\rangle)\right) \\ 
& = \sqrt{\hbar/2m\omega}(\alpha^{*}+\alpha) = \sqrt{2\hbar/m\omega}\Re(\alpha) , \\
\langle X^{2}\rangle_{\alpha} & = (\hbar/2m\omega)\langle\alpha|(a^{\dagger}+a)^{2}|\alpha\rangle = (\hbar/2m\omega)\langle\alpha|((a^{\dagger})^{2}+a^{\dagger}a+aa^{\dagger}+a^{2})|\alpha\rangle \\ 
& = (\hbar/2m\omega)\langle\alpha|((a^{\dagger})^{2}+a^{\dagger}a+a^{\dagger}a+\mathbb{I}+a^{2})|\alpha\rangle \\
& = (\hbar/2m\omega)((\alpha^{*})^{2}+2|\alpha|^{2}+1+\alpha^{2}), \\
Var(X)_{\alpha} & = \langle X^{2}\rangle_{\alpha} - \langle X\rangle_{\alpha}^{2} \\ 
& = (\hbar/2m\omega)[(\alpha^{*})^{2}+2|\alpha|^{2}+1+\alpha^{2} - (\alpha^{*})^{2}-2|\alpha|^{2}-\alpha^{2}] \\ 
& = \hbar/2m\omega.
\end{align}
Para o momento linear teremos:
\begin{align}
\langle P\rangle_{\alpha} & = i\sqrt{\hbar m\omega/2}(\alpha^{*}-\alpha) = \sqrt{2\hbar m\omega}\Im(\alpha), \\
\langle P^{2}\rangle_{\alpha} & = -(\hbar m\omega/2)((\alpha^{*})^{2}-2|\alpha|^{2}-1+\alpha^{2}), \\
Var(P)_{\alpha} & = \hbar m\omega/2.
\end{align}
Vemos assim que
\begin{equation}
Var(X)_{\alpha}Var(P)_{\alpha}=\frac{\hbar^{2}}{4},
\end{equation}
que é o _mínimo_ possível para a _relação de incerteza_ de Heisenberg, que veremos mais adiante, para qualquer que seja $\alpha$.

### Dinâmica do estado coerente sob $H_{ohq}$
Vamos considerar um oscilador preparado inicialmente em um estado coerente, $|\psi_{0}\rangle=|\alpha\rangle$, e evoluindo sob a ação do Hamiltoniano do OHQ, $U_{t}=e^{-i\hbar\omega(N+\mathbb{I}/2)t/\hbar} = e^{-i\omega t\mathbb{I}/2}e^{-i\omega tN}\equiv e^{-i\omega t/2}e^{-i\omega tN}$. O estado desse sistema é obtido como segue
\begin{align}
|\psi_{t}\rangle & = U_{t}|\alpha\rangle = e^{-i\omega t/2}e^{-i\omega tN}e^{-|\alpha|^{2}/2}\sum_{n=0}^{\infty}\frac{\alpha^{n}}{\sqrt{n!}}|n\rangle = e^{-i\omega t/2}e^{-|\alpha|^{2}/2}\sum_{n=0}^{\infty}\frac{\alpha^{n}}{\sqrt{n!}}e^{-i\omega tN}|n\rangle \\
& = e^{-i\omega t/2}e^{-|\alpha|^{2}/2}\sum_{n=0}^{\infty}\frac{\alpha^{n}}{\sqrt{n!}}e^{-i\omega tn}|n\rangle = e^{-i\omega t/2}e^{-|\alpha|^{2}/2}\sum_{n=0}^{\infty}\frac{\left(\alpha e^{-i\omega t}\right)^{n}}{\sqrt{n!}}|n\rangle \\
& = e^{-i\omega t/2}e^{-\left|\alpha e^{-i\omega t}\right|^{2}/2}\sum_{n=0}^{\infty}\frac{\left(\alpha e^{-i\omega t}\right)^{n}}{\sqrt{n!}}|n\rangle \equiv e^{-\left|\alpha_{t}\right|^{2}/2}\sum_{n=0}^{\infty}\frac{\alpha_{t}^{n}}{\sqrt{n!}}|n\rangle \\
& = |\alpha_{t}\rangle,
\end{align}
com $\alpha_{t}:=\alpha e^{-i\omega t}$. Podemos ver que essa dinâmica corresponde a um giro no plano complexo:
\begin{cases}
\Re(\alpha_{t}) = \Re(\alpha e^{-i\omega t}) = \Re(\alpha)\Re(e^{-i\omega t}) - \Im(\alpha)\Im(e^{-i\omega t}) = \cos(\omega t)\Re(\alpha) -(-\sin(\omega t))\Im(\alpha), \\
\Im(\alpha_{t}) = \Im(\alpha e^{-i\omega t}) = \Re(\alpha)\Im(e^{-i\omega t}) + \Im(\alpha)\Re(e^{-i\omega t}) = -\sin(\omega t)\Re(\alpha) +\cos(\omega t)\Im(\alpha),
\end{cases}
ou
\begin{equation}
\begin{bmatrix} \Re(\alpha_{t}) \\ \Im(\alpha_{t}) \end{bmatrix} = \begin{bmatrix} \cos(\omega t) & \sin(\omega t) \\ -\sin(\omega t) & \cos(\omega t) \end{bmatrix}\begin{bmatrix} \Re(\alpha) \\ \Im(\alpha) \end{bmatrix}.
\end{equation}

Ademais, como o estado evoluído é um estado coerente, teremos
\begin{align}
\langle X\rangle_{\alpha_{t}} & =  \sqrt{\hbar/2m\omega}(\alpha^{*}_{t}+\alpha_{t}) = \sqrt{2\hbar/m\omega}\Re(\alpha_{t}), \\ 
\langle P\rangle_{\alpha_{t}} & = i\sqrt{\hbar m\omega/2}(\alpha^{*}_{t}-\alpha_{t}) = \sqrt{2\hbar m\omega}\Im(\alpha_{t}).
\end{align}
Portanto, a dinâmica de $\left(\sqrt{2m\omega/\hbar}\langle X\rangle_{\alpha_{t}},\sqrt{1/2\hbar m\omega}\langle P\rangle_{\alpha_{t}}\right)$ é a mesma que aquela mostrada na figura abaixo. Esse resultado é esperado em vista dos resultados obtidos acima na representação de Heisenberg.

```python
def alpha_t(Re_a,Im_a,tmax):
    fig = plt.figure(figsize=(5,5)); w = 1; t = np.arange(0,tmax,0.01)
    x = np.cos(w*t)*Re_a + np.sin(w*t)*Im_a; y = -np.sin(w*t)*Re_a + np.cos(w*t)*Im_a
    plt.plot(x,y,linestyle='dashed'); plt.xlim(-1,1); plt.ylim(-1,1)
    plt.xlabel(r'$\Re(\alpha_{t})$',fontsize=13); plt.ylabel(r'$\Im(\alpha_{t})$',fontsize=13)
    plt.annotate(r'$t=$'+str(tmax)[:5], xy=(0.6,0.8), xytext=(0.6,0.8), fontsize=12)
    xm = cos(w*tmax)*Re_a + sin(w*tmax)*Im_a; ym = -sin(w*tmax)*Re_a + cos(w*tmax)*Im_a
    plt.annotate(r'$\alpha_{t}$', xy=(xm,ym), xytext=(xm,ym), fontsize=12); plt.show()
interactive(alpha_t, Re_a = (0,1,0.1), Im_a = (0,1,0.1), tmax = (0,20,0.05))
```

### Estados coerentes formam uma base super-completa
Como vocês verificarão nos exercícios, estados coerentes diferentes não são ortonormais. Aqui verificaremos que, além disso, eles formam uma base supercompleta, o que significa que se somarmos os projetores em todos os estados coerentes obteremos um operador maior que o operador identidade. Mais explicitamente, verificaremos que
\begin{equation}
\iint|\alpha\rangle\langle\alpha|d^{2}\alpha=\pi\mathbb{I}.
\end{equation}
Para fazer a integral sobre o plano complexo, usamos
\begin{equation}
\alpha=|\alpha|e^{i\theta} \text{ e } d^{2}\alpha=|\alpha|d\theta d|\alpha|.
\end{equation}
Assim
\begin{align}
\iint|\alpha\rangle\langle\alpha|d^{2}\alpha & = \int_{0}^{\infty}d|\alpha|\int_{0}^{2\pi}d\theta|\alpha|e^{-|\alpha|^{2}}\sum_{n=0}^{\infty}\frac{|\alpha|^{n}e^{i\theta n}}{\sqrt{n!}}|n\rangle\sum_{m=0}^{\infty}\frac{|\alpha|^{m}e^{-i\theta m}}{\sqrt{m!}}\langle m|\\
 & = \sum_{n,m=0}^{\infty}\frac{1}{\sqrt{n!m!}}|n\rangle\langle m|\int_{0}^{\infty}|\alpha|^{n+m+1}e^{-|\alpha|^{2}}d|\alpha|\stackrel{=2\pi\delta_{nm}}{\overbrace{\int_{0}^{2\pi}e^{i\theta(n-m)}d\theta}}\\
 & = 2\pi\sum_{n=0}^{\infty}\frac{1}{n!}|n\rangle\langle n|\int_{0}^{\infty}|\alpha|^{2n+1}e^{-|\alpha|^{2}}d|\alpha|.
\end{align}
Fazendo a mudança de variável $x=|\alpha|^{2}\Rightarrow dx=2|\alpha|d|\alpha|$ teremos (notamos que $|\alpha|^{2n+1}=(|\alpha|^{2})^{n}|\alpha|$)
\begin{align}
\int|\alpha\rangle\langle\alpha|d^{2}\alpha & = 2\pi\sum_{n=0}^{\infty}\frac{1}{n!}|n\rangle\langle n|\int_{0}^{\infty}x^{n}|\alpha|e^{-x}\frac{dx}{2|\alpha|} \\ 
& =\pi\sum_{n=0}^{\infty}\frac{1}{n!}|n\rangle\langle n|\int_{0}^{\infty}x^{n}e^{-x}dx=\pi\sum_{n=0}^{\infty}\frac{1}{n!}|n\rangle\langle n|\Gamma(n+1)\\
 & = \pi\sum_{n=0}^{\infty}\frac{1}{n!}|n\rangle\langle n|n!=\pi\mathbb{I}.
\end{align}
Para mais detalhes sobre a função gama, veja https://github.com/jonasmaziero/mecanica_quantica_1900-1925/blob/master/01_planck.ipynb.

# Oscilador quântico forçado
Consideremos um oscilador sujeito a uma força de restauração e a uma outra força constante no espaço mas dependente do tempo
\begin{equation}
f_{t}\therefore U=-\int_{0}^{x} f_{t}dx'=-f_{t}x.
\end{equation}
O Hamiltoniano quantizado fica escrito como
\begin{align}
H_{t} &= \frac{P^{2}}{2m}+\frac{m\omega^{2}X^{2}}{2} -f_{t}X \\ 
& =: \hbar\omega(a^{\dagger}a+\mathbb{I}/2)-F_{t}(a+a^{\dagger}),
\end{align}
onde definimos
\begin{equation}
F_{t}:=f_{t}\sqrt{\hbar/2m\omega}.
\end{equation}
Como vocês verificarão,
\begin{equation}
[H_{t},H_{t'}]\ne 0 \text{ se } F_{t}\ne F_{t'}.
\end{equation}
Então, a princípio, nesse caso teríamos que recorrer às séries de Dyson. No entando, podemos obter algumas informações por outro caminho (veja a Ref. 3).

### Dinâmica na representacão de Heisenberg
Consideremos a eq. de Heisenberg
\begin{align}
i\hbar\partial_{t}a_{H} & = [a_{H},H_{t}^{H}] = \left[a_{H},\hbar\omega(a_{H}^{\dagger}a_{H}+\mathbb{I}/2)-F_{t}(a_{H}+a_{H}^{\dagger})\right] \\
& = \hbar\omega([a_{H},a_{H}^{\dagger}a_{H}]+0) - F_{t}([a_{H},a_{H}]+[a_{H},a_{H}^{\dagger}]) \\
& = \hbar\omega a_{H} - F_{t}\mathbb{I}, \\
\therefore \partial_{t}a_{H} & = -i\omega a_{H} + (iF_{t}/\hbar)\mathbb{I}.
\end{align}

Notemos que
\begin{align}
& \partial_{t}\left(e^{i\omega t}a_{H}\right) = \left(\partial_{t}e^{i\omega t}\right)a_{H} + e^{i\omega t}\partial_{t}a_{H} = i\omega e^{i\omega t}a_{H} + e^{i\omega t}\partial_{t}a_{H}, \\
& \therefore e^{-i\omega t}\partial_{t}\left(e^{i\omega t}a_{H}\right) = i\omega a_{H} + \partial_{t}a_{H} = (iF_{t}/\hbar)\mathbb{I}, \\
& \therefore \partial_{t}\left(e^{i\omega t}a_{H}\right) = e^{i\omega t}(iF_{t}/\hbar)\mathbb{I}, \\
& \therefore \int_{0}^{t}dt'\partial_{t'}\left(e^{i\omega t'}a_{H}(t')\right) = \int_{0}^{t}dt'e^{i\omega t'}(iF_{t'}/\hbar)\mathbb{I}, \\
& \therefore e^{i\omega t}a_{H}(t)-e^{i\omega 0}a_{H}(0) = \frac{i}{\hbar}\int_{0}^{t}dt'e^{i\omega t'}F_{t'}\mathbb{I}, \\
& \therefore a_{H} = e^{-i\omega t}(a + v_{t}\mathbb{I}) \text{ e } a_{H}^{\dagger} = e^{i\omega t}(a^{\dagger} + v_{t}^{*}\mathbb{I}),
\end{align}
onde definimos
\begin{equation}
v_{t} := \frac{i}{\hbar}\int_{0}^{t}dt'e^{i\omega t'}F_{t'} = \frac{i}{\sqrt{2\hbar m\omega}} \int_{0}^{t}dt'e^{i\omega t'}f_{t'}.
\end{equation}
Note que $v_{t}:\mathbb{R}\rightarrow\mathbb{C}$.

Conhecendo $a_{H}$ podemos escrever a dependência de muitos observáveis, com por exemplo
\begin{align}
X_{t}^{H} & = \sqrt{\frac{\hbar}{2m\omega}}(a_{H}^{\dagger}+a_{H}) = \sqrt{\frac{\hbar}{2m\omega}}\left(e^{i\omega t}(a^{\dagger} + v_{t}^{*}\mathbb{I}) + e^{-i\omega t}(a + v_{t}\mathbb{I})\right) \\
& = \sqrt{\frac{\hbar}{2m\omega}}\left((\cos(\omega t)+i\sin(\omega t))(a^{\dagger} + v_{t}^{*}\mathbb{I}) + (\cos(\omega t)-i\sin(\omega t))(a + v_{t}\mathbb{I})\right)\\
& = \sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}+a)\cos(\omega t) + i\sqrt{\frac{\hbar}{2m\omega}}(a^{\dagger}-a)\sin(\omega t)\frac{m\omega}{m\omega} \\
& \hspace{0.5cm}+\sqrt{\frac{\hbar}{2m\omega}}\cos(\omega t)(v_{t}^{*}+v_{t})\mathbb{I} + i\sqrt{\frac{\hbar}{2m\omega}}\sin(\omega t)(v_{t}^{*}-v_{t})\mathbb{I} \\
& = X\cos\omega t + \frac{P}{m\omega}\sin\omega t +2\sqrt{\frac{\hbar}{2m\omega}}\left(\Re(v_{t})\cos\omega t + \Im(v_{t})\sin\omega t\right)\mathbb{I}, \\
P_{t}^{H} & = i\sqrt{\frac{\hbar m\omega}{2}}(a_{H}^{\dagger}-a_{H}) = i\sqrt{\frac{\hbar m\omega}{2}}\left(e^{i\omega t}(a^{\dagger} + v_{t}^{*}\mathbb{I}) - e^{-i\omega t}(a + v_{t}\mathbb{I})\right), \\
& \vdots \\
\frac{P_{t}^{H}}{m\omega} &= -X\sin\omega t + \frac{P}{m\omega}\cos\omega t + 2\sqrt{\frac{\hbar}{2m\omega}}\left(\Im(v_{t})\cos\omega t - \Re(v_{t})\sin\omega t\right)\mathbb{I}.
\end{align}
Na forma matricial teremos
\begin{equation}
\begin{bmatrix} X_{t}^{H} \\ \frac{P_{t}^{H}}{m\omega} \end{bmatrix} = \begin{bmatrix} \cos\omega t & \sin\omega t \\ -\sin\omega t & \cos\omega t \end{bmatrix}\begin{bmatrix} X + \sqrt{\frac{2\hbar}{m\omega}}\Re(v_{t})\mathbb{I} \\ \frac{P}{m\omega} + \sqrt{\frac{2\hbar}{m\omega}}\Im(v_{t})\mathbb{I} \end{bmatrix}.
\end{equation}

Nesse caso, a distância do ponto $\left(\langle X_{t}^{H}\rangle,\langle P_{t}^{H}\rangle/m\omega\right)$ à origem $(0,0)$ é
\begin{equation}
\sqrt{\langle X\rangle^{2}+\langle P/m\omega\rangle^{2} + \sqrt{\frac{\hbar}{2m\omega}}\left(\langle X\rangle\Re(v_{t})+\langle P/m\omega\rangle\Im(v_{t})\right) +\frac{2\hbar}{m\omega}|v_{t}|^{2} }.
\end{equation}

Alguns casos particulares:
* Força constante $f_{t}=f$:
\begin{align}
v_{t} & = \frac{i}{\sqrt{2\hbar m\omega}} \int_{0}^{t}dt'e^{i\omega t'}f_{t'} = \frac{if}{\sqrt{2\hbar m\omega}}\left( \int_{0}^{t}dt'\cos(\omega t')+i\int_{0}^{t}dt'\sin(\omega t')\right) \\
& = \frac{if}{\sqrt{2\hbar m\omega}\omega}\left(\left(\sin(\omega t')\right)_{0}^{t}-i\left(\cos(\omega t')\right)_{0}^{t}\right) \\
& = \frac{if}{\sqrt{2\hbar m\omega}\omega}\left(\left(\sin(\omega t)-\sin(0)\right)-i\left(\cos(\omega t)-\cos(0)\right)\right) \\
& = \frac{f}{\sqrt{2\hbar m\omega}\omega}\left(i\sin(\omega t)+\left(\cos(\omega t)-1\right)\right).
\end{align}

* Força oscilante, senoidal, $f_{t}=f\sin(\omega_{f}t)$:
\begin{align}
v_{t} & = \frac{i}{\sqrt{2\hbar m\omega}} \int_{0}^{t}dt'e^{i\omega t'}f\sin(\omega_{f}t') = \frac{if}{\sqrt{2\hbar m\omega}} \int_{0}^{t}dt'e^{i\omega t'}\frac{e^{i\omega_{f}t'}-e^{-i\omega_{f}t'}}{2i} \\
& = \frac{f}{2\sqrt{2\hbar m\omega}}\left(\int_{0}^{t}dt'e^{i(\omega+\omega_{f})t'}-\int_{0}^{t}dt'e^{i(\omega-\omega_{f})t'}\right) = \frac{f}{\sqrt{8\hbar m\omega}}\left(\frac{e^{i(\omega+\omega_{f})t}-1}{i(\omega+\omega_{f})}-\frac{e^{i(\omega-\omega_{f})t}-1}{i(\omega-\omega_{f})}\right) \\
& = \cdots \\
& = \frac{f}{\sqrt{2\hbar m\omega}}\frac{i\left(\omega\sin(\omega t)\sin(\omega_{f}t)+\omega_{f}\cos(\omega t)\cos(\omega_{f}t)\right) +\left(\omega\cos(\omega t)\sin(\omega_{f}t)-\omega_{f}\sin(\omega t)\cos(\omega_{f}t)\right)}{\omega^{2}-\omega_{f}^{2}}.
\end{align}

```python
# Animação da dinâmica do oscilador forçado no "espaço de fases"
def oqf_dyn(x0,p0,tmax,wf):
    fig = plt.figure(figsize=(9,6)); w = 1; m = 1; f = 0.3; hbb = 1
    t = np.arange(0,tmax,0.01)
    ctt = math.sqrt((2*hbb)/(m*w)); comax = math.cos(w*tmax); simax = math.sin(w*tmax)
    co = np.cos(w*t); si = np.sin(w*t); 
    caso = 'osc'
    if caso == 'cte': # pra força constante
        vtr = (f/math.sqrt(2*hbb*m*w**3))*(co-1); vti = (f/math.sqrt(2*hbb*m*w**3))*si 
        vtrmax = (f/math.sqrt(2*hbb*m*w**3))*(comax-1); vtimax = (f/math.sqrt(2*hbb*m*w**3))*simax
    elif caso == 'osc': # pra força oscilante
        cof = np.cos(wf*t); sif = np.sin(wf*t); cofmax = math.cos(wf*tmax); sifmax = math.sin(wf*tmax)
        cter = f/((w**2-wf**2)*math.sqrt(2*hbb*m*w))
        vtr = (w*co*sif - wf*si*cof)*cter; vti = (w*si*sif - wf*co*cof)*cter
        vtrmax = (w*comax*sifmax - wf*simax*cofmax)*cter; vtimax = (w*simax*sifmax - wf*comax*cofmax)*cter
    x = co*(x0 + ctt*vtr) + si*(p0 + ctt*vti); y = -si*(x0 + ctt*vtr) + co*(p0 + ctt*vti)
    xm = comax*(x0 + ctt*vtrmax) + simax*(p0 + ctt*vtimax); ym = -simax*(x0 + ctt*vtrmax) + comax*(p0 + ctt*vtimax) # para o ponto atual
    plt.plot(x,y, linestyle='dashed'); plt.xlim(-3,3); plt.ylim(-2,2)
    plt.xlabel(r'$\langle X_{t}^{H}\rangle$',fontsize=13)
    plt.ylabel(r'$\langle P_{t}^{H}\rangle/m\omega$',fontsize=13)
    plt.annotate(r'$t=$'+str(tmax), xy=(2,1.8), xytext=(2,1.8), fontsize=12)
    plt.annotate(r'$\bigcirc$', xy=(xm,ym), xytext=(xm,ym), fontsize=12)
    plt.show()
interactive(oqf_dyn, x0 = (0,1,0.1), p0 = (0,1,0.1), tmax = (0,100,0.2), wf = (0.51,1.51,0.025))
```

### Dinâmica na representação de Schrödinger (para $|\psi_{0}\rangle=|0\rangle$)
Embora não tenhamos resolvido a eq. de Schrödinger para $U_{t}$, podemos usar as equações obtidas acima para mostrar como o estado fundamental do oscilador evolui quando submetido à força $f_{t}$:
\begin{align}
& a_{H}|0\rangle = e^{-i\omega t}(a + v_{t}\mathbb{I})|0\rangle, \\
& \therefore U_{t}^{\dagger}aU_{t}|0\rangle = e^{-i\omega t}(a|0\rangle + v_{t}\mathbb{I}|0\rangle)=e^{-i\omega t}(|\oslash\rangle + v_{t}|0\rangle) = e^{-i\omega t}v_{t}|0\rangle, \\
& \therefore U_{t}U_{t}^{\dagger}aU_{t}|0\rangle = a(U_{t}|0\rangle) = e^{-i\omega t}v_{t}(U_{t}|0\rangle).
\end{align}
Vemos assim que o estado evoluído no tempo $U_{t}|0\rangle$ é autovetor do operador de abaixamento $a$, e que portanto é um estado coerente
\begin{equation}
U_{t}|0\rangle=|\alpha_{t}\rangle \text{ com } \alpha_{t} = e^{-i\omega t}v_{t}.
\end{equation}
