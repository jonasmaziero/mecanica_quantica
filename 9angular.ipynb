# Momento angular
Quantizando a expressão clássica
\begin{align}
\vec{L} & = \vec{X}\mathrm{x}\vec{P} \\
        & = \sum_{j=1}^{3}\hat{e}_{j}(\vec{X}\mathrm{x}\vec{P})_{j} \\
        & = \sum_{j=1}^{3}\hat{e}_{j}\sum_{k,l=1}^{3}\varepsilon_{jkl}X_{k}P_{l} \\ 
        & = \sum_{j=1}^{3}\hat{e}_{j}L_{j}.
\end{align}

## Relação de comutação para as componentes do momento angular
Usando novamente $[A,BC]=B[A,C]+[A,B]C$ e $[AB,C] = A[B,C]+[A,C]B$, teremos
\begin{align}
[L_{j},L_{k}] & = \left[\sum_{l,m=1}^{3}\varepsilon_{jlm}X_{l}P_{m} , \sum_{p,q=1}^{3}\varepsilon_{kpq}X_{p}P_{q} \right] \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left[X_{l}P_{m} , X_{p}P_{q} \right] \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(X_{l}[P_{m} , X_{p}P_{q}] + [X_{l} , X_{p}P_{q}]P_{m} \right) \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(X_{l}X_{p}[P_{m},P_{q}] + X_{l}[P_{m},X_{p}]P_{q} + X_{p}[X_{l},P_{q}]P_{m} + [X_{l} , X_{p}]P_{q}P_{m} \right) \\
              & = \sum_{l,m,p,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kpq}\left(\mathbb{0} + X_{l}(-i\hbar\mathbb{1}\delta_{mp})P_{q} + X_{p}(i\hbar\mathbb{1}\delta_{lq})P_{m} + \mathbb{0} \right) \\
              & = -i\hbar\sum_{l,m,q=1}^{3}\varepsilon_{jlm}\varepsilon_{kmq}X_{l}P_{q} + i\hbar\sum_{l,m,p=1}^{3}\varepsilon_{jlm}\varepsilon_{kpl}X_{p}P_{m} \\
\end{align}
Agora usamos
\begin{equation}
\sum_{j=1}^{3}\varepsilon_{jkl}\varepsilon_{jmn} = \delta_{km}\delta_{ln} - \delta_{kn}\delta_{lm}
\end{equation}
para obter
\begin{align}
[L_{j},L_{k}] & = i\hbar\sum_{l,q=1}^{3}\sum_{m=1}^{3}\varepsilon_{mjl}\varepsilon_{mkq}X_{l}P_{q} - i\hbar\sum_{m,p=1}^{3}\sum_{l=1}^{3}\varepsilon_{ljm}\varepsilon_{lkp}X_{p}P_{m} \\
              & = i\hbar\sum_{l,q=1}^{3}(\delta_{jk}\delta_{lq}-\delta_{jq}\delta_{lk})X_{l}P_{q} - i\hbar\sum_{m,p=1}^{3}(\delta_{jk}\delta_{mp}-\delta_{jp}\delta_{mk})X_{p}P_{m}.
\end{align}
Como para $j=k$ temos $[L_{j},L_{k}]=\mathbb{0}$, vamos considerar a relação acima com $j\ne k\Rightarrow \delta_{jk}=0$. Assim
\begin{align}
[L_{j},L_{k}] & = -i\hbar\sum_{l,q=1}^{3}\delta_{jq}\delta_{lk}X_{l}P_{q} + i\hbar\sum_{m,p=1}^{3}\delta_{jp}\delta_{mk}X_{p}P_{m} \\
              & = i\hbar(-X_{k}P_{j} + X_{j}P_{k}).
\end{align}
Substituindo os índices,
\begin{align}
[L_{1},L_{2}] & = i\hbar(X_{1}P_{2}-X_{2}P_{1}) = i\hbar\sum_{j,k}\varepsilon_{3jk}X_{j}P_{k} = i\hbar L_{3} =  i\hbar\varepsilon_{123}L_{3}, \\
[L_{2},L_{3}] & = i\hbar(X_{2}P_{3}-X_{3}P_{2}) = i\hbar\sum_{j,k}\varepsilon_{1jk}X_{j}P_{k} = i\hbar L_{1} = i\hbar\varepsilon_{231}L_{1}, \\
[L_{1},L_{3}] & = i\hbar(X_{1}P_{3}-X_{3}P_{1}) = -i\hbar\sum_{j,k}\varepsilon_{2jk}X_{j}P_{k} = -i\hbar L_{2} = i\hbar\varepsilon_{132}L_{2}. \\
\end{align}
Finalmente a importante relação de comutação para as componentes do momento angular escrita em uma forma geral:
\begin{equation}
[L_{j},L_{k}] = i\hbar\varepsilon_{jkl}L_{l} = i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l}.
\end{equation}

## Autovalores do momento angular
O método que usaremos aqui é semelhante aquele utilizado na solução do oscilador harmônico. Primeiramente verifiquemos que o quadrado do momento angular comuta com todas as suas componentes:
\begin{align}
[L^{2},L_{k}] & = \left[\sum_{j=1}^{3}L_{j}^{2},L_{k}\right] \\
              & = \sum_{j=1}^{3}[L_{j}L_{j},L_{k}] \\
              & = \sum_{j=1}^{3}(L_{j}[L_{j},L_{k}] + [L_{j},L_{k}]L_{j}) \\
              & = \sum_{j=1}^{3}(L_{j}i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l} + i\hbar\sum_{l=1}^{3}\varepsilon_{jkl}L_{l}L_{j}) \\
              & = i\hbar\sum_{j,l=1}^{3}\varepsilon_{jkl}(L_{j}L_{l} + L_{l}L_{j}) \\
              & = i\hbar\sum_{j,l=1}^{3}\varepsilon_{jkl}\{L_{j},L_{l}\}.
\end{align}
Considerando as componentes explicitamente,
\begin{align}
[L^{2},L_{1}] & = i\hbar(\varepsilon_{213}\{L_{2},L_{3}\} + \varepsilon_{312}\{L_{3},L_{2}\}) \\ 
              & = i\hbar\{L_{2},L_{3}\}(\varepsilon_{213} + \varepsilon_{312}) \\ 
              & = i\hbar\{L_{2},L_{3}\}(\varepsilon_{213} - \varepsilon_{312}) \\ 
              & = \mathbb{0},
\end{align}
onde usamos $\{L_{j},L_{k}\}=\{L_{k},L_{j}\}$. <br>
__Exercício:__ Faça a verificação análoga para $k=1$ e $k=2$. <br>
Com isso vemos que
\begin{equation}
[L^{2},L_{k}] = \mathbb{0}
\end{equation}
para $k=1,2,3$.

seguindo, consideremos que $L^{2}$ compartilha a base de autovetores $|l,m\rangle$ com $L_{3}$ e
\begin{align}
L^{2}|l,m\rangle & = \lambda|l,m\rangle \\
L_{3}|l,m\rangle & = \hbar m|l,m\rangle.
\end{align}
Vamos verificar que $\lambda = \hbar^{2}l(l+1)$ com $l=n/2$ para $n\in\mathbb{N}$ e $m=-l,-l+1,\cdots,0\cdots,l-1,l$. Para isso precisaremos dos _operadores escada_:
\begin{equation}
L_{\pm} = L_{1} \pm iL_{2}.
\end{equation}
Como
\begin{equation}
[L^{2},L_{\pm}] = [L^{2},L_{1}] \pm i[L^{2},L_{2}] = \mathbb{0}, 
\end{equation}
vemos que os vetores $L_{\pm}|l,m\rangle$ também são autovetores de $L^{2}$ correspondentes aos mesmos autovalores $\lambda$ que $|l,m\rangle$. Por outro lado,
\begin{align}
[L_{3},L_{\pm}] & = [L_{3},L_{1}] \pm i[L_{3},L_{2}] \\
                & = i\hbar\varepsilon_{312}L_{2} \pm i i\hbar\varepsilon_{321}L_{1} \\
                & = i\hbar L_{2} \pm \hbar L_{1} \\
                & = \pm\hbar(L_{1} \pm iL_{2}) \\
                & = \pm\hbar L_{\pm}
\end{align}
e
\begin{align}
L_{3}(L_{\pm}|l,m\rangle) & = (L_{3}L_{\pm})|l,m\rangle \\
                          & = (\pm\hbar L_{\pm} + L_{\pm}L_{3})|l,m\rangle \\
                          & = \pm\hbar L_{\pm}|l,m\rangle + L_{\pm}L_{3}|l,m\rangle \\
                          & = \pm\hbar L_{\pm}|l,m\rangle + L_{\pm}\hbar m|l,m\rangle \\
                          & = \hbar( m \pm 1)(L_{\pm}|l,m\rangle).
\end{align}
Ou seja, os vetores $L_{\pm}|l,m\rangle$ também são autovetores de $L_{3}$, mas com autovalores aumentados ou diminuídos de $\hbar$ em relação àquele associado com $|l,m\rangle$. Agora, como $L^{2} = L_{1}^{2}+L_{2}^{2}+L_{3}^{2}$, não moderemos aumentar nem diminuir indefinidamente os autovalores de $L_{3}$ através da aplicação dos operadores escada. Por isso, para os valores definidos por $m=\pm l$ a ação dos operadores escada nos retorna o vetor nulo:
\begin{equation}
L_{\pm}|l,m =\pm l\rangle = |\oslash\rangle.
\end{equation}

Continuando, consideremos
\begin{align}
L_{\pm}L_{\mp} & = (L_{1} \pm iL_{2})(L_{1} \mp iL_{2}) \\
               & = L_{1}^{2} \mp iL_{1}L_{2} \pm iL_{2}L_{1} + L_{2}^{2} \\
               & = L_{1}^{2} + L_{2}^{2} + L_{3}^{2} - L_{3}^{2} \mp i[L_{1},L_{2}]  \\
               & = L^{2} - L_{3}^{2} \mp ii\hbar\varepsilon_{123}L_{3} \\
               & = L^{2} - L_{3}^{2} \pm\hbar L_{3}.
\end{align}
Então, podemos considerar
\begin{align}
L^{2}|l,l\rangle & = (L_{-}L_{+} + L_{3}^{2} + \hbar L_{3})|l,l\rangle \\
                 & = L_{-}L_{+}|l,l\rangle + L_{3}^{2}|l,l\rangle + \hbar L_{3}|l,l\rangle \\
                 & = |\oslash\rangle + \hbar^{2}l^{2}|l,l\rangle + \hbar\hbar l|l,l\rangle \\
                 & = \hbar^{2}l(l+1)|l,l\rangle \\
                 & = \lambda|l,l\rangle.
\end{align}

Para obter o domínio de $l$, consideremos que os autovalores de $L_{3}$ são, por definição, $\hbar m$ e que vimos que se começamos no estado $|l,m=-l\rangle$ e aplicamos $L_{+}$ $n$ vezes, aumentaremos o valor de $m$ de uma unidade em cada aplicação de $L_{+}$, até obtermos o estado $|l,m=l\rangle$. Como $n\in\mathbb{N}_{>0}$ teremos que $l=-l+n\Rightarrow 2l=n \Rightarrow l = n/2$.
