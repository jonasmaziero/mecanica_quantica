# Sistemas compostos
O caminho moderno da descrição quântica da Natureza é começando com sistemas simples e adicionando complexidade aos poucos. Vimos como estados de um único grau de liberdade são descritos. Vejamos agora como fica o postulado dos estados quando o sistema que queremos descrever é composto por dois ou mais graus de liberdade.

## Postulado dos estados
O estado de um sistema composto pelos graus de liberdade enumerados por $s=1,2,\cdots,n$ é descrito por um vetor no espaço de Hilbert obtido tomando-se o produto tensorial dos espaços de Hilbert individuais:
\begin{equation}
|\Psi\rangle \in \mathcal{H} = \mathcal{H}_{1}\otimes\mathcal{H}_{2}\otimes\cdots\otimes\mathcal{H}_{n}=\bigotimes_{s=1}^{n}\mathcal{H}_{s}.
\end{equation}

_OBS:_ $\dim\mathcal{H}=:d = d_{1}d_{2}\cdots d_{n} = \Pi_{s=1}^{n}d_{s}$, com $d_{s}\dim\mathcal{H}_{s}$.

O produto tensorial de duas matrizes quaisquer $A\in\mathbb{C}^{m\mathrm{x}n}$  e $B\in\mathbb{C}^{p\mathrm{x}q}$ é uma matriz $C\in\mathbb{C}^{mp\mathrm{x}nq}$ definida como:
\begin{align}
C & = A\otimes B \\ 
  & = \begin{bmatrix} A_{11} & A_{12} & \cdots & A_{1n} \\
                A_{21} & A_{22} & \cdots & A_{2n} \\
                \vdots & \vdots & \cdots & \vdots \\
                A_{m1} & A_{m2} & \cdots & A_{mn}
      \end{bmatrix}
\otimes
      \begin{bmatrix} B_{11} & B_{12} & \cdots & B_{1q} \\
                B_{21} & B_{22} & \cdots & B_{2q} \\
                \vdots & \vdots & \cdots & \vdots \\
                B_{p1} & B_{p2} & \cdots & B_{pq}
      \end{bmatrix}
      \\
& = \begin{bmatrix} A_{11}B & A_{12}B & \cdots & A_{1n}B \\
                    A_{21}B & A_{22}B & \cdots & A_{2n}B \\
                    \vdots & \vdots & \cdots & \vdots \\
                    A_{m1}B & A_{m2}B & \cdots & A_{mn}B
      \end{bmatrix}.
\end{align}

Note que se conhecemos as bases locais $\{|\psi_{j}^{(s)}\rangle \}_{j=1}^{d_{s}}$, obteremos uma base para o espaço total usando o conjunto de todos os vetores:
\begin{equation}
\{|\psi_{j_{1}}\rangle\otimes|\psi_{j_{2}}\rangle\cdots\otimes|\psi_{j_{n}}\rangle\} = \left\{\bigotimes_{s=1}^{n}|\psi_{j_{s}}\rangle\right\}.
\end{equation}
Qualquer vetor de $\mathcal{H}$ pode ser expresso como
\begin{equation}
|\Psi\rangle = \sum_{j_{1}=1}^{d_{1}}\sum_{j_{2}=1}^{d_{2}}\cdots\sum_{j_{n}=1}^{d_{n}} c_{j_{1}j_{2}\cdots j_{n}} |\psi_{j_{1}}\rangle\otimes|\psi_{j_{2}}\rangle\cdots\otimes|\psi_{j_{n}}\rangle,
\end{equation}
com
\begin{equation}
c_{j_{1}j_{2}\cdots j_{n}} = \left(\langle\psi_{j_{1}}|\otimes\langle\psi_{j_{2}}|\cdots\otimes\langle\psi_{j_{n}}|\right)|\Psi\rangle.
\end{equation}

_OBS:_ Observáveis de um sistema composto são descritos por matrizes Hermitianas de dimensão $d$. Se queremos nos referir a um observável do sistema $1$, por exemplo, escrevemos $O_{s}\otimes\mathbb{I}_{2}\otimes\cdots\otimes\mathbb{I}_{n}$.

_OBS:_ É frequente o uso das igualdades
\begin{align}
& (A\otimes B)(C\otimes D)=AC\otimes BD, \\
& (A\otimes B)^{\dagger}=A^{\dagger}\otimes B^{\dagger}.
\end{align}

### Exemplo: Dois sistemas de dois níveis
Aqui o espaço de estados é $\mathbb{C}^{4} = \mathbb{C}^{2}\otimes\mathbb{C}^{2}$. Os estados puros desse sistema são quaisquer vetores normalizados de $\mathbb{C}^{4}$. Os operadores densidade serão quaisquer matrizes de $\mathbb{C}^{4x4}$ positivas-semidefinidas e de traço igual a um.

Vamos considerar um sistema de dois níveis em um estado qualquer desconhecido $|?\rangle$. Digamos que medimos $\sigma_{z}$ em cada um dos sistemas, mas separadamente. Se os resultado $z_{+}$ é obtido no sistema $1$ e $z_{-}$ é obtido no sistema $2$, o estado preparado será:
\begin{equation}
|z_{+}\rangle\otimes|z_{-}\rangle= \begin{bmatrix} 0 \\ 1 \\ 0 \\ 0 \end{bmatrix}
\end{equation}
na base $\{|z_{+}\rangle\otimes|z_{+}\rangle, |z_{+}\rangle\otimes|z_{-}\rangle, |z_{-}\rangle\otimes|z_{+}\rangle, |z_{-}\rangle\otimes|z_{-}\rangle\}$.
Vamos considerar que esses sistemas são dipolos magnéticos, cujo Hamiltoniano de interação é
\begin{equation}
H = (D/2)[(\vec{\sigma}\otimes\sigma_{0})\cdot(\sigma_{0}\otimes\vec{\sigma})-3\hat{n}\cdot\vec{\sigma}\otimes\hat{n}\cdot\vec{\sigma}],
\end{equation}
com $D=\mu_{0}\gamma^{2}/16\pi r^{3}$.

Para $\hat{n} = (0,0,1)$, usamos $\vec{\sigma}=\sum_{j=1}^{3}\sigma_{j}\hat{e}_{j}$ para obter (__exercício__)
\begin{align}
H & = (D/2)(\sigma_{1}\otimes\sigma_{1}+\sigma_{2}\otimes\sigma_{2}-2\sigma_{3}\otimes\sigma_{3}) \\
  & = 0|\Psi_{-}\rangle\langle\Psi_{-}|+2D|\Psi_{+}\rangle\langle\Psi_{+}|-D(|\Phi_{-}\rangle\langle\Phi_{-}|+|\Phi_{+}\rangle\langle\Phi_{+}|),
\end{align}
com $|\Psi_{\pm}\rangle=2^{-1/2}(|z_{+}\rangle\otimes|z_{-}\rangle\pm|z_{-}\rangle\otimes|z_{+}\rangle)$ e $|\Phi_{\pm}\rangle=2^{-1/2}(|z_{+}\rangle\otimes|z_{+}\rangle\pm|z_{-}\rangle\otimes|z_{-}\rangle)$. <br>
_OBS:_ Para diagonalizar esse Hamiltoniano, use as funções do arquivo jupyterQ.ipynb, que pode ser obtido em https://github.com/jonasmaziero/jupyterQ. No novo arquivo importe essas funções usando o comando %run 'jupyterQ.ipynb'.

Se os dipolos interagem por um tempo $t=\pi/4D$ obteremos o estado (__exercício:__ veja e.g. https://arxiv.org/abs/1805.01518):
\begin{align}
|\Psi_{t}\rangle & = U_{t}|z_{+}\rangle\otimes|z_{-}\rangle = e^{-iHt}|z_{+}\rangle\otimes|z_{-}\rangle \\ 
                 & = \left(|z_{+}\rangle\otimes|z_{-}\rangle - i|z_{-}\rangle\otimes|z_{+}\rangle\right)/\sqrt{2}.
\end{align}
Este é um estado emaranhado, que é uma superposição envolvendo estados ortogonais de duas ou mais partículas e que não pode ser escrita como um produto tensorial de vetores dos espaços individuais, i.e., $|\Psi_{t}\rangle\ne|\omega_{1}\rangle\otimes|\delta_{2}\rangle$.

## Regra de Born para sistemas compostos
Se medimos os observáveis $O^{(s)}=\sum_{j=1}^{d_{s}}o_{j}^{(s)}|o_{j}^{(s)}\rangle\langle o_{j}^{(s)}|$ para $s=1,2$, teremos a seguinte generalização direta para a regra de Born:
\begin{equation}
Pr\left(o_{j}^{(1)},o_{k}^{(2)}|\Psi\right) = \left|\langle\Psi|\left(|o_{j}^{(1)}\rangle\otimes|o_{k}^{(2)}\rangle\right)\right|^{2}.
\end{equation}

Podemos obter tal generalização no caso em que somente um dos sistemas é medido da seguinte forma. Digamos que para um sistema bipartido preparado no estado $|\Psi\rangle$ queremos obter as probabilidades para medidas de $O^{(1)}$. Neste caso, fazemos de conta que $O^{(2)}$ também foi medido e consideramos a probabilidade marginal:
\begin{equation}
Pr\left(o_{j}^{(1)}|\Psi\right) = \sum_{k}Pr\left(o_{j}^{(1)},o_{k}^{(2)}|\Psi\right).
\end{equation}

## Estado pós-medida para sistemas compostos
No caso de um único grau de liberdade, pela _repetibilidade_, sabemos que o estado do sistema pós-medida quando o valor obtido na medida do observável $O$ for $o_{j}$ é o autovetor correspondente $|o_{j}\rangle$. Podemos escrever isso de outra maneira. Podemos dizer que obter $o_{j}$ projeta o estado do sistema na direção de $|o_{j}\rangle$. Assim o estado normalizado pós-medida poder ser escrito como
\begin{equation}
|\psi\rangle \xrightarrow[]{o_{j}} \frac{\Pi_{o_{j}}|\psi\rangle}{||\Pi_{o_{j}}|\psi\rangle||} = \frac{|o_{j}\rangle\langle o_{j}|\psi\rangle}{|||o_{j}\rangle\langle o_{j}|\psi\rangle||} = |o_{j}\rangle.
\end{equation}

Para sistemas compostos, se medimos observáveis de todos os subsistemas, a generalização é
\begin{equation}
|\Psi\rangle \xrightarrow[]{o_{j_{1}},\cdots,o_{j_{n}}} \frac{\Pi_{o_{j_{1}}}\otimes\cdots\otimes\Pi_{o_{j_{n}}}|\Psi\rangle}{||\Pi_{o_{j_{1}}}\otimes\cdots\otimes\Pi_{o_{j_{n}}}|\Psi\rangle||}.
\end{equation}
Se, por outro lado, fazemos medidas e.g. somente no subsistema $1$, teremos o estado pós-medida
\begin{equation}
|\Psi\rangle \xrightarrow[]{o_{j_{1}}} \frac{\Pi_{o_{j_{1}}}\otimes\mathbb{I}_{2}\otimes\cdots\otimes\mathbb{I}_{n}|\Psi\rangle}{||\Pi_{o_{j_{1}}}\otimes\mathbb{I}_{2}\otimes\cdots\otimes\mathbb{I}_{n}|\Psi\rangle||}.
\end{equation}
A extensão direta dessa ideia possibilita a obtenção do estado para medidas em alguns dos subsistemas.

## O paradoxo EPR
Em 1935 Einstein, Podolski e Rosen propuseram um experimento de pensamento com o intuito de mostrar que a MQ não era uma teoria completa, que pela definição deles seria uma teoria com a qual podemos prever com certeza o resultado de experimentos antes de fazê-los e que naturalmente é local conforme estabelecido pela teoria da relatividade. O argumento de EPR pode ser descrito como segue. Consideremos o seguinte estado de duas partículas de spin 1/2:
\begin{equation}
|\Psi_{-}\rangle = \frac{1}{\sqrt{2}}\left(|z_{+}\rangle\otimes|z_{-}\rangle - |z_{-}\rangle\otimes|z_{+}\rangle\right).
\end{equation}
Usando $|x_{\pm}\rangle=(|z_{+}\rangle\pm|z_{-}\rangle)/\sqrt{2}$ podemos verificar que (__exercício__):
\begin{equation}
|\Psi_{-}\rangle = \frac{e^{i\pi}}{\sqrt{2}}\left(|x_{+}\rangle\otimes|x_{-}\rangle - |x_{-}\rangle\otimes|x_{+}\rangle\right).
\end{equation}
Agora considere que medimos um dos observáveis $\sigma_{z}$ ou $\sigma_{x}$ da partícula da "esquerda". Se $z_{\pm}$ ou $x_{\pm}$ é obtido nessas medidas, pela correlação que há entre os spins, sabemos que o estado da partícula da direita depois da medida será $|z_{\mp}\rangle$ ou $|x_{\mp}\rangle$. Agora, se impomos _localidade_, ou seja, que as medidas feitas na esquerda não podem afetar o resultado das medidas que foram, que serão ou que estão sendo feitas na direita (espacialmente distante), vemos que pelo uso do estado emaranhado acima podemos prever, com certeza e sem perturbar o sistema, os resultados de medidas de dois _observáveis incompatíveis_ $\sigma_{z}$ ou $\sigma_{x}$ da partícula da "direita". Como isso é claramente impossível segundo as regras da MQ, EPR disseram que embora a MQ estivesse correta, eles acreditavam na existência de uma teoria completa que a substituiria.

## O teorema de Bell
Em 1964 John Bell, que trabalhava no CERN, tentou mostrar que EPR estavam certos, mas com uma abordagem que possibilitaría fazer testes experimentais. Bell supôs que existiriam variáveis ocultadas, tais que se conhecêssemos seus valores, ou distribuição de probabilidades desses valores, poderíamos prever com certeza os resultados de qualquer medida em MQ. Essa suposição pode ser chamada de _determinismo_. Bell também assumiu _localidade_, ou seja, que o resultado da medida em um sistema não pode influenciar os resultados de medidas no outro sistema. Agora, consideramos o seguinte observável
\begin{align}
O & = A_{1}B_{1} + A_{1}B_{2} + A_{2}B_{1} - A_{2}B_{2} \\
  & = A_{1}(B_{1} + B_{2}) + A_{2}(B_{1} - B_{2}),
\end{align}
com $A_{j}$ e $B_{k}$ sendo observáveis, todos tendo espectro $\pm 1$, dos subsistemas $A$ e $B$, respectivamente. Segundo as suposições acima, o valor médio de $O$ ficaria
\begin{equation}
\langle O\rangle_{dl} = \int d\lambda\rho(\lambda) \left\{a_{1}(\lambda)[b_{1}(\lambda) + b_{2}(\lambda)] + a_{2}(\lambda)[b_{1}(\lambda) - b_{2}(\lambda)]\right\},
\end{equation}
em que denotamos por $a_{j},b_{k}$ os possíveis valores de $A_{j},B_{k}$ e $\rho(\lambda)$ é a distribuição de probabilidades para as variáveis ocultas. Note que a suposição de determinismo entra na dependência dos valores dos observáveis com $\lambda$ e a localidade entra na não dependência desses valores com aqueles do outro sistema. O contrário seria e.g. $a_{j}=a_{j}(\lambda,b_{k})$. Considerando todas as combidacões possíveveis de valores de $A_{j}$ e $B_{k}$ vemos que o espectro de $O$ é (__exercício__) $\pm2$. Disso segue que o valor médio desse observável deve estar limitado por
\begin{equation}
-2 \le \langle O\rangle_{dl} \le 2.
\end{equation}

Agora consideremos um situação idêntica mas descrita pelas regras da MQ. Usaremos como estado preparado $|\Psi_{-}\rangle$. Quanto aos observáveis, uma escolha apropriada é $A_{1}=\sigma_{z}$, $A_{2}=\sigma_{x}$, $B_{1}=(\sigma_{z}+\sigma_{x})/\sqrt{2}$ e $B_{2}=(\sigma_{z}-\sigma_{x})/\sqrt{2}$. Segue que (__exercício__) $O=\sqrt{2}(\sigma_{z}\otimes\sigma_{z}+\sigma_{x}\otimes\sigma_{x})$, $\sigma_{z}\otimes\sigma_{z}|\Psi_{-}\rangle=-|\Psi_{-}\rangle$ e $\sigma_{x}\otimes\sigma_{x}|\Psi_{-}\rangle=-|\Psi_{-}\rangle$. Assim,
\begin{align}
\langle O\rangle_{mq} & = \langle\Psi_{-}|O|\Psi_{-}\rangle \\
                      & = \langle\Psi_{-}|\sqrt{2}(-2|\Psi_{-}\rangle) \\
                      & = -2\sqrt{2} < -2.
\end{align}
Ou seja, o resultado da MQ está fora do limite imposto pela teoria de variáveis ocultas locais. Desde 1982, vários experimentos foram feitos fornecendo resultados compatíveis com a MQ. Isso nos mostra que uma ou as duas suposições estão erradas, i.e., a Natureza é não determinística e/ou não local. Bell preferia esta última opção, pois existe uma versão da MQ, a de de Broglie-Bohm, que é determinística e não local. 

## Comunicação superluminal?
Esses resultados motivaram propostas engenhosas para fazer comunicação mais rápido que a luz. Uma proposta foi a seguinte. Considere que duas partes, chamadas usualmente de Alice e Bob querem trocar bits de informação e que eles compartilham pares de partículas de spin 1/2 preparadas no estado $|\Psi_{-}\rangle$. Alice e Bob combinam que se Alice quer mandar o bit $0$ ela mede o observável $\sigma_{z}$ no seu spin do par e se quiser mandar o bit $1$ ela mede $\sigma_{x}$. Por sua vez, Bob faz várias cópias do estado do seu spin de cada par e mede o observável $\sigma_{z}$. Se Alice mediu $\sigma_{z}$ a distribuição de probabilidades obtida nas medidas de Bob, $\{Pr(z_{+}),Pr(z_{-})\}$, será (__exercício__) $\{1,0\}$ ou $\{0,1\}$; se Alice no entanto mediu $\sigma_{x}$, Bob obterá, aproximadamente, a DP  (__exercício__) $\{1/2,1/2\}$. Com isso, se a distância entre eles for adequada, eles conseguem trocar bits de informação mais rápido do que a luz demoraria para ir da posição de Alice até Bob. Ou seja, eles fazem comunicação superluminal, voltam no tempo, e .... <br>

## Teorema da não clonagem em MQ
O problema com a proposta anterior é que não existe Xerox quântico. Uma prova simplificada disto é a seguinte. Considera que existe uma dinâmica dada pelo operador de evolução temporal $U$ que atua no espaço $\mathcal{H}_{s}\otimes\mathcal{H}_{a}$, com $s$ sendo o sistema cujo estado queremos copiar e $a$ é um sistema auxiliar no qual  copiaremos aquele estado. Consideremos que $U$ pode copiar dois estados desconhecidos quaisquer $|\psi\rangle$ e $|\phi\rangle$, i.e.,
\begin{align}
U|\psi\rangle\otimes|\alpha\rangle & = |\psi\rangle\otimes|\psi\rangle, \\
U|\phi\rangle\otimes|\alpha\rangle & = |\phi\rangle\otimes|\phi\rangle,
\end{align}
com $|\alpha\rangle$ sendo o estado inicial do sistema auxiliar. Segue dessas relações que
\begin{align}
\langle\psi|\otimes\langle\alpha|U^{\dagger}U|\phi\rangle\otimes|\alpha\rangle & = (\langle\psi|\otimes\langle\psi|)(|\phi\rangle\otimes|\phi\rangle) \\
\langle\psi|\phi\rangle\otimes\langle\alpha|\alpha\rangle & = \langle\psi|\phi\rangle\otimes\langle\psi|\phi\rangle \\
\langle\psi|\phi\rangle & = \langle\psi|\phi\rangle^{2}.
\end{align}
Ou seja, $\langle\psi|\phi\rangle$ é igual a $0$ ou $1$. Ou seja, só poderá existir uma máquina de copiar estados desconhecidos $U$ se os estados forem iguais ou ortogonais (caso clássico).

Vale observar que máquinas de copiar estados quânticos de forma imperfeita ou probabilística são factíveis, mas não são "boas o suficiente" para fazer comunicação superluminal [https://arxiv.org/abs/quant-ph/0511088].
